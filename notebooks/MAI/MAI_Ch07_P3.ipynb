### Resolução

#### a) Calcule as reações nos apoios A e G. <a id="A"></a> ([_enunciatum_](#PIII))

**Grau de hiperestaticidade**

Comece-se por analisar a estrutura observando que que esta treliça possui 4 ligações ao exterior (4 reações: 1 em A, 1 em B e 2 em G) pelo que, $r > 3$ (treliça exteriormente hiperestática do 1º grau).

**Estatia global**

Estatia global da treliça ($\alpha_g$) é calculada de acordo com:

\begin{equation*}
\alpha_g = \alpha_e + \alpha_i
\end{equation*}

em que $\alpha_e$ e $\alpha_i$ são a estatia exterior e estatia interior da treliça calculadas, respectivamente, por:

\begin{equation*}
\alpha_e = 3 × \textrm{( nº de apoios − 1)} − \textrm{nº de libertações exteriores}
\end{equation*}

- Apoios: O número de apoios refere-se ao total de apoios que a treliça tem
- Libertações Exteriores: Este termo refere-se ao número de graus de liberdade que não estão restringidos pelos apoios

\begin{equation*}
\alpha_i = 3 × \textrm{nº de malhas fechadas} − \textrm{nº de ligações interiores}
\end{equation*}

- Malha fechada numa treliça refere-se a uma região da estrutura que é completamente delimitada por barras, formando um circuito fechado. Malhas fechadas considera que cada malha adiciona três graus de liberdade internos.
- Ligações interiores referem-se às barras internas da treliça que conectam os nós entre si.


```python
def estatiaexterior(napoios, nlibertacoes):
    return 3 * ( napoios - 1 ) - nlibertacoes

def estatiainterior(nmalhas,nligainter):
    return  3 * nmalhas - nligainter

npoios_, nlibext_ = 3, 5
alphe = estatiaexterior(npoios_, nlibext_)
print(f'estatia exterior = 3({npoios_} - 1) - {nlibext_} =  {alphe}')
nmalh_, nlibint_ = 4, 1*2 + 2*5 
alphi = estatiainterior(nmalh_,nlibint_)
print(f'estatia interior = 3({nmalh_}) - {nlibint_} = {alphi}')
estatiaglobal = alphe + alphi
print((f'estatia global = {estatiaglobal}'))
```

**Equilíbrio da treliça - determinação das reações**

Por conseguinte, para calcular todas as reações procede-se à separação da treliça em duas partes conforme se mostra pela figura (i.1) e (1.2) relativas à parte esquerda e direita da treliça, respectivamente.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch07/P3/MAI_Ch07_P3a.jpg"
width="800">

A análise do equilíbrio da treliça plana conduz ao seguinte sistema de equações lineares:

_Estrutura da Esquerda (Figura i.1)_

\begin{equation*}
\begin{cases}
\rightarrow(+)&:&\sum F_x = 0 \\
\uparrow(+)&:&\sum F_y = 0 \\
\circlearrowleft(+)&:&\sum M_D =0
\end{cases}~\Leftrightarrow~
\begin{cases}
-R_{Bx} + F_h + R_{Dx} &=~0 \\
R_{Ay} - F_v - R_{Dy} &=~0 \\
-R_{Ay}(2L_x) + F_h(L_y) &=~0
\end{cases}
\end{equation*}

com $L_x = 1$ m; $L_y = 4$ m.

_Estrutura da Direita (Figura i.2)_

\begin{equation*}
\begin{cases}
\rightarrow(+)&:&\sum F_x = 0 \\
\uparrow(+)&:&\sum F_y = 0 \\
\circlearrowleft(+)&:&\sum M_D =0
\end{cases}~\Leftrightarrow~
\begin{cases}
R_{Gx} - R_{Dx} &=~0 \\
R_{Dy} - R_{Gy} &=~0 \\
-R_{Gy}(2L_x) + R_{Gx} (L_y) &=~0
\end{cases}
\end{equation*}

Resolvendo o sistema de equações em conjunto:

\begin{equation*}
\Leftrightarrow~
\begin{cases}
R_{Bx} &=~9~\textrm{kN} \\
R_{Dy} &=~8~\textrm{kN} \\
R_{Ay} &=~10~\textrm{kN} \\
R_{Dx} &=~4~\textrm{kN} \\
R_{Gy} &=~8~\textrm{kN} \\
R_{Gx} &=~4~\textrm{kN} \\
\end{cases}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import solve

# unit: m, kN
LX, LY, FV, FH = 1., 4., 2., 5.

rbx,ray,rdx,rdy,rgx,rgy,fv,fh,a,b = \
    sy.symbols('rbx ray rdx rdy rgx rgy fv fh a b')

print('-------------------')
print('Sistema de Equações')
print('-------------------')

SFX_e = -rbx + fh + rdx
print('sum Fx :', SFX_e)
SFY_e = ray - fv - rdy
print('sum Fy :', SFY_e)
SMD_e = -2*a*ray + fh*b
print('sum MD :', SMD_e)

SFX_d = rgx - rdx
print('sum Fx :', SFX_d)
SFY_d = rdy - rgy
print('sum Fy :', SFY_d)
SMD_d = -2*a*rgy + b*rgx
print('sum MD :', SMD_d)
```

```python

sol = solve({SFX_e,SFY_e,SMD_e,SFX_d,SFY_d,SMD_d},{rbx,ray,rdx,rdy,rgx,rgy})

print('------')
print('Reações')
print('-------')

RBx = sol[rbx]
NRBx = sy.N(RBx.subs({(fv,FV),(fh,FH),(a,LX),(b,LY)}))
print(f'RBx = {RBx} = {NRBx:.1f} kN')

RAy = sol[ray]
NRAy = sy.N(RAy.subs({(fv,FV),(fh,FH),(a,LX),(b,LY)}))
print(f'RAy = {RAy}       = {NRAy:.1f} kN')

RDx = sol[rdx]
NRDx = sy.N(RDx.subs({(fv,FV),(fh,FH),(a,LX),(b,LY)}))
print(f'RDx = {RDx}   = {NRDx:.1f} kN')

RDy = sol[rdy]
NRDy = sy.N(RDy.subs({(fv,FV),(fh,FH),(a,LX),(b,LY)}))
print(f'RDy = {RDy} = {NRDy:.1f} kN')

RGx = sol[rgx]
NRGx = sy.N(RGx.subs({(fv,FV),(fh,FH),(a,LX),(b,LY)}))
print(f'RGx = {RGx}   = {NRGx:.1f} kN')

RGy = sol[rgy]
NRGy = sy.N(RGy.subs({(fv,FV),(fh,FH),(a,LX),(b,LY)}))
print(f'RGy = {RGy} = {NRGy:.1f} kN')
```

#### b) Calcule as forças nas barras da treliça indicando se são de tração ou compressão. <a id="B"></a> ([_enunciatum_](#PIII))

**Método dos Nós**

**Nó A: Escolha do primeiro nó**

Comece-se por escolher um nó com apenas duas incógnitas. Procede-se então ao equilíbrio estático do nó A.

_Diagrama de corpo livre do nó A_

Considere-se o DCL do nó A, em que:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch07/P3/MAI_Ch07_P3noA.jpg"
width="300">

em que,

\begin{equation*}
\alpha = \arctan\left(\frac{L_y}{2L_x}\right) = \arctan\left
(\frac{4}{2}\right) =
63,43^\circ
\end{equation*}

```python
import numpy as np

alpha = np.arctan(LY/2/LX)
print(f'alpha (nó A) = {alpha:.3f} radianos = {np.rad2deg(alpha):.2f} graus')
```

Em conformidade com a terceira lei de Newton que afirma que, as forças de ação e reação entre corpos em contacto têm a mesma intensidade, a mesma linha de acção, e sentidos opostos, forças com sentido para fora dos nós indicam tração das barras sobre esse nó, e que, inversamente, forças que entram no nó, indicam que a barra associadas está à compressão.

_Estudo do equilíbrio estático no nó_

Aplicando as equações de equilíbrio ao nó resulta:

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
F_{AC}\cos\alpha + F_h &=~0 \\
F_{AC}\sin\alpha + F_{BA} + R_{Ay} &=~0
\end{cases}~\Leftrightarrow~
\begin{cases}
F_{AC} &= -\frac{F_h}{\cos(\alpha)} = -11.18 ~\textrm{kN (C)}\\
F_{BA} &= F_h \tan(\alpha) - R_{Ay} =  0 ~\textrm{kN}
\end{cases}
\end{equation*}

```python
fac, fba, t = sy.symbols('fac, fba t')

print('-------------------')
print('Sistema de Equações')
print('-------------------')

SFX = fac*sy.cos(t) + fh
print('sum Fx :', SFX)
SFY = fac*sy.sin(t) + fba + ray
print('sum Fy :', SFY)
sol = solve({SFX,SFY},{fac, fba})

print('-------------------')
print('Forças nas barras')
print('-------------------')

FAC = sol[fac]
NFAC = sy.N(FAC.subs({(t,alpha),(ray,NRAy),(fh,FH)}))
print(f'FAC = {FAC} = {NFAC:.2f} kN')

FBA = sol[fba]
NFBA = np.abs(sy.N(FBA.subs({(t,alpha),(ray,NRAy),(fh,FH)})))
print(f'FBA = {FBA} = {NFBA:.2f} kN')
```

**Nó B**

Escolhe-se um outro nó onde sejam desconhecidas apenas duas forças. Faz-se agora o equilíbrio do nó B conforme se mostra pela figura abaixo em que:

_Diagrama de corpo livre_

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch07/P3/MAI_Ch07_P3noB.jpg"
width="300">

em que,

\begin{equation*}
\beta = \arctan\left(\frac{L_y/2}{L_x}\right) = \arctan\left
(\frac{2}{1}\right) =
63,43^\circ (\equiv \alpha)
\end{equation*}

```python
beta = np.arctan(LY/2/LX)
print(f'beta (nó B) = {beta:.3f} radianos = {np.rad2deg(beta):.2f} graus')
```

_Equilíbrio estático_

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
-R_{Bx} + F_{FD} + F_{CB}\cos\beta = 0 \\
F_{CB}\sin\beta + F_{BA} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
F_{BD} = R_{Bx} = 9~\textrm{kN (T)} \\
F_{CB} = 0~\textrm{kN}
\end{cases}
\end{equation*}

```python
fbd, fcb = sy.symbols('fbd fcb')

print('-------------------')
print('Sistema de Equações')
print('-------------------')

SFX = -rbx + fbd + fcb*sy.cos(t)
print('sum Fx :', SFX)
SFY = fcb*sy.sin(beta) + fba
print('sum Fy :', SFY)
sol = solve({SFX,SFY},{fbd,fcb})

print('-------------------')
print('Forças nas barras')
print('-------------------')

FBD = sol[fbd]
NFBD = FBD.subs({(t,beta),(fba,NFBA),(rbx,NRBx)})
print(f'FBD = {FBD} = {NFBD:.2f} kN (T)')

FCB = sol[fcb]
NFCB = np.abs(sy.N(FCB.subs({(t,beta),(fba,NFBA),(rbx,NRBx)})))
print(f'FCB = {FCB} = {NFCB:.2f} kN')
```

Note-se que sendo a força $F_{BC}$ nula, olhando para o nó C (vide figura abaixo) conclui-se que as forças $F_{AC}$ e $F_{CD}$ terão necessariamente que igualar-se

**Nó G**

Do equilíbrio do nó G (veja-se o DCL na figura abaixo), obtêm-se as forças $F_{GE}$ e $F_{GF}$.

_Diagrama de corpo livre_

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch07/P3/MAI_Ch07_P3noG.jpg"
width="300">

_Equilíbrio estático_

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
R_{Gx} - F_{GE}\cos\alpha = 0 \\
- R_{Gy} + F_{GF} + F_{GE}\sin\alpha = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
F_{GE} = \frac{R_{GX}}{\cos\alpha} = 8.94~\textrm{kN (T)}\\
F_{GF} = -R_{GX}\tan\alpha + R_{GY} = 0~\textrm{kN}
\end{cases}
\end{equation*}

```python
fge, fgf = sy.symbols('fge fgf')

print('-------------------')
print('Sistema de Equações')
print('-------------------')

SFX = rgx - fge*sy.cos(t)
print('sum Fx :', SFX)
SFY = - rgy + fgf + fge*sy.sin(t)
print('sum Fy :', SFY)
sol = solve({SFX,SFY},{fge, fgf})

print('-------------------')
print('Forças nas barras')
print('-------------------')

FGE = sol[fge]
NFGE = sy.N(FGE.subs({(rgx,NRGx),(rgy,NRGy),(t,alpha)}))
print(f'FGE = {FGE} = {NFGE:.2f} kN (T)')

FGF = sol[fgf]
NFGF = sy.N(FGF.subs({(rgx,NRGx),(rgy,NRGy),(t,alpha)}))
print(f'FGF = {FGF} = {NFGF:.2f} kN')
```

**Nó F**

Finalmente no nó F, o DCL da figura abaixo permite estabelecer as equações de equilíbrio estático seguintes:

_Diagrama de corpo livre_


<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch07/P3/MAI_Ch07_P3noF.jpg"
width="300">

_Equilíbrio estático_

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
F_{FD} - F_{FE}\cos\beta = 0 \\
F_{FG} - F_{FE}\sin\beta = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
F_{FD} = \frac{F_{GF}}{\tan\alpha} = 0~\textrm{kN}\\
F_{FE} = \frac{F_{GF}}{\sin\alpha} = 0~\textrm{kN}
\end{cases}
\end{equation*}

```python
ffg = fgf
ffd, ffe = sy.symbols('ffd ffe')

print('-------------------')
print('Sistema de Equações')
print('-------------------')

SFX = ffd - ffe*sy.cos(t)
print('sum Fx :', SFX)
SFY = ffg - ffe*sy.sin(t)
print('sum Fy :', SFY)
sol = solve({SFX,SFY},{ffd, ffe})

print('-------------------')
print('Forças nas barras')
print('-------------------')

FFD = sol[ffd]
NFFD = sy.N(FFD.subs({(ffg,NFGF),(rgy,NRGy),(t,beta)}))
print(f'FFD = {FFD} = {NFFD:.2f} kN')

FFE = sol[ffe]
NFFE = sy.N(FFE.subs({(ffg,NFGF),(rgy,NRGy),(t,beta)}))
print(f'FFE = {FFE} = {NFFE:.2f} kN')
```

Analogamente ao nó C, uma vez que a força $F_{FE}$ é nula, e olhando para o nó E (veja-se a figura abaixo) conclui-se também que as forças $F_{GE}$ e $F_{ED}$ são iguais.
