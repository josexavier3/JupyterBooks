### Problema 4

Determine a força horizontal força $C_x$, que o pino C exerce na barra BC por forma a que o mecanismo mostrado na figura se encontre em equilíbrio quando $\theta = 45^\circ$. Despreze o efeito da massa de cada uma das barras.

### Resolução

A reação $C_x$, no apoio em C, pode ser obtida pela libertação da restrição de movimentar na direção $x$, permitindo que a estrutura se desloque nessa direção. O sistema tem apenas um grau de liberdade, definido pela coordenada independente $\theta$. Quando $\theta$ sofre uma rotação virtual positiva $\delta\theta$, apenas a força redundante $C_x$ e a força vertical $P$ realizam trabalho exterior.

#### Diagrama de Corpo Livre

Pelo princípio da sobreposição, considera-se um sistema mecânico equivalente com um apoio duplo em A e um apoio simples em C, onde a estrutura equivalente está submetida ao carregamento real ($P$) mais uma carregamento redundante que fisicamente representa a reação horizontal em C, $C_x$.

Considere-se um sistema de coordenadas com origem em A. Assuma-se um deslocamento virtual consistente com um deslocamento horizontal do ponto C: 

- As forças de reação em A, $A_x,A_y$, não exercem trabalho virtual uma vez que os deslocamentos virtuais dos seus pontos de aplicação são nulos. 
- Considera-se desprezável o peso próprio da estrutura face ao carregamento externo a que está submetido. 
- Desta forma, as forças externas aplicadas à estrutura que realizam trabalho virtual são a força vertical em B e a componente da força de reação $C_x$. De notar que uma vez que o deslocamento virtual no ponto C é horizontal, a componente $C_y$ não exercerá trabalho virtual ($\cos 90^\circ = 0$).

#### Equações do Princípio dos Trabalhos Virtuais

Por aplicação do PTV:

\begin{equation*}
\delta U = \delta U_F + \delta U_{C_x} = 0
~\Leftrightarrow~
\vec{F}\cdot\delta\vec{y}_B +
\vec{C_x}\cdot\delta\vec{x}_C = 0
\end{equation*}

em que,

\begin{equation*}
\vec{F} = -200 \,\vec{\jmath}
\quad\wedge\quad
\vec{C_x} = C_x \,\vec{\imath}
\end{equation*}


#### Deslocamentos virtuais (pontos de aplicação das forças)

- Deslocamento virtual $y_B$

Por relações geométricas é fácil concluir que:

\begin{equation*}
y_B = AB\sin\theta
\end{equation*}

considerando uma variação infinitesimal dos deslocamentos e rotações virtuais:

\begin{equation*}
\delta y_D = AB\cos\theta\delta\theta
\end{equation*}

ou seja,

\begin{equation*}
\delta\vec{y}_B = (AB\cos\theta\delta\theta)\,\vec{\jmath}
\end{equation*}

- Deslocamento virtual $x_C$

Para relacionar a coordenada $x_C$ com a varíavel angular $\theta$, considere-se a seguinte expressão trigonométrica que expressa a lei dos cosenos:

\begin{equation*}
BC^2 = AB^2 + AC^2 - 2AB\cdot AC\cdot\cos\theta
~\Leftrightarrow~
BC^2 = AB^2 + x_C^2 - 2AB\cdot x_C\cdot\cos\theta
\end{equation*}

Considere-se agora a variação infinitesimal dos deslocamento virtuais $\delta x_C$ e $\delta \theta$:

\begin{equation*}
\begin{aligned}
&0 = 0 + 2x_C\delta x_C - 2AB (\delta x_C \cos\theta - x_C\sin\theta\delta\theta)\\
~\Leftrightarrow~&
\delta x_C = -\frac{AB x_C \sin\theta}{x_C-AB\cos\theta}\,\delta\theta
\end{aligned}
\end{equation*}

ou seja,

\begin{equation*}
\delta\vec{x}_C =
\left(-\frac{AB x_C \sin\theta}{x_C-AB\cos\theta}\,\delta\theta \right)\,\vec{\imath}
\end{equation*}


Desenvolvendo o PTV:

\begin{equation*}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
-F\cdot \delta y_C - C_x\cdot \delta x_C = 0\\
&~\Leftrightarrow~
-F\cdot (AB\cos\theta\delta\theta)
- C_x\cdot \left(-\frac{AB x_C \sin\theta}{x_C-AB\cos\theta}\,\delta\theta \right) = 0 \\
&~\Leftrightarrow~
-F\cdot (AB\cos\theta)
+ C_x\cdot \left(\frac{AB x_C \sin\theta}{x_C-AB\cos\theta} \right) = 0\\
&~\Leftrightarrow~
C_x = \frac{x_C-AB\cos\theta}{x_C\tan\theta}F
\end{aligned}
\end{equation*}

De notar que, quando $\theta = 45^\circ = \frac{\sqrt{2}}{2}$ o valor de $x_C$ vem da expressão anterior:

\begin{equation*}
\begin{aligned}
&x_C^2 - (2AB\cos\theta)\,x_C + (AB^2 -BC^2) = 0\\
\Leftrightarrow~&
x_C^2 - (\sqrt{2}AB)\,x_C + (AB^2 -BC^2) = 0 \\
\Leftrightarrow~&
x_C = \frac{AB \pm \sqrt{(2 BC^2-AB^2)}}{\sqrt{2}}
\end{aligned}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
from typing import NamedTuple

# Dados
# unidades SI: m, N, deg
class dados(NamedTuple):
    AB : float 
    BC : float
    F : float
    ang : float

var = dados(0.6, 0.7, 200., 45.)
print(f'dados = {var}')

f, ab, bc, t, cx, xc = sy.symbols('f ab bc t cx xc')

EQ = -f*ab*sy.cos(t) + cx*(ab*xc*sy.sin(t)/(xc-ab*sy.cos(t)))
sol1 = solve({EQ},{cx})
CX = sol1[cx]
print(' Cx = ',sy.simplify(CX))

theta_rad = np.deg2rad(var.ang) # unit: radianos

EX = xc**2 - (2*var.AB*np.cos(theta_rad))*xc + (var.AB**2 - var.BC**2)
sol2 = solve({EX},{xc})
XC = sol2[1][xc]
# NXC = XC.subs({(ab,AB),(bc,BC),(t,theta)})

NCX = CX.subs({(f,var.F),(t,theta_rad),(xc,XC),(ab,var.AB)})
print(f'para theta = {var.ang:.0f} deg')
print(f' xC({var.ang:.0f}º) = {XC:.3f} m')
print(f' Cx({var.ang:.0f}º) = {NCX:.1f} N')
```

#### Solução alternativa

\begin{equation*}
x_C = AB\cos\theta + BC \cos\beta
\end{equation*}

Considerando a variação infinitéssimal das grandezas:

\begin{equation*}
\delta x_C = -AB\sin\theta\delta\theta - BC \sin\beta\delta\beta
\end{equation*}

A relação entre os ângulos $\theta$ e $\beta$ é expressa por:

\begin{equation*}
\sin\beta = \frac{AB}{BC}\sin\theta
\end{equation*}

Por conseguinte, a variação do ângulo $\theta$ e $\beta$ é:

\begin{equation*}
\cos\beta\delta\beta = \frac{AB}{BC}\cos\theta\delta\theta
\end{equation*}

de onde resulta,

\begin{equation*}
\delta\beta = \frac{AB}{BC}\frac{\cos\theta}{\cos\beta}\delta\theta
\end{equation*}

Exprimindo agora $\delta x_C$ apenas em função da variação infinitesimal $\delta\theta$

\begin{equation*}
\delta x_C = -AB\sin\theta\delta\theta - AB\sin\beta\left(\frac{AB}{BC}\frac{\cos\theta}{\cos\beta}\delta\theta\right)
\end{equation*}

simplificando,

\begin{equation*}
\delta x_C  = -AB [\sin\theta + \tan\beta\cos\theta]\delta\theta 
\end{equation*}
 
O PTV vem assim:

\begin{equation*}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
-F\cdot \delta y_C - C_x\cdot \delta x_C = 0\\
&~\Leftrightarrow~
-F\cdot (AB\cos\theta\delta\theta)
- C_x\cdot (-AB [\sin\theta + \tan\beta\cos\theta])\delta\theta = 0 \\
&~\Leftrightarrow~
-F\cdot (AB\cos\theta)
+ C_x\cdot (AB [\sin\theta + \tan\beta\cos\theta]) = 0\\
&~\Leftrightarrow~
C_x = \frac{\cos\theta}{\sin\theta + \tan\beta\cos\theta}F
\end{aligned}
\end{equation*}

```python
beta_ang = np.arcsin(var.AB / var.BC * np.sin(theta_rad))
print(f'para theta = {var.ang:.0f} deg')
print(f'beta = {np.rad2deg(beta_ang):.1f} deg')
xc_2 = var.AB * np.cos(theta_rad) + var.BC * np.cos(beta_ang)
CX_2 = np.cos(theta_rad) / (np.sin(theta_rad) + np.tan(beta_ang)*np.cos(theta_rad)) * var.F
print(f' xC({var.ang:.0f}º) = {xc_2:.3f} m')
print(f' Cx({var.ang:.0f}º) = {CX_2:.1f} N')
```
