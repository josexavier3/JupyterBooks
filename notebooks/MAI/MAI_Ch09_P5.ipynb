#### Resolução

**Diagrama de corpo livre + Configuração virtual**

Considere-se o diagrama de corpo livre do mecanismo da figura enunciada, representando neste todas as forças e momentos atuantes. Admita uma perturbação da configuração inicial pela imposição de uma  rotação virtual $\delta\theta$ à barra EC. Ao fazê-lo, e atendendo a que a configuração virtual deverá ser cinematicamente admissível, percebe-se que a barra AB tenderá a rodar também de $\delta\theta$ conforme, aliás, se ilustra pela figura. É de notar que, em consequência, a barra BC não roda no plano e tem apenas um movimento de translação.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P5/MAI_Ch09_P5a.jpg"
style="max-width: 100%;"/>

**Que forças/momentos realizam trabalho no movimento virtual considerado?**

- Na configuração virtual, as forças de reação $A_x$, $A_y$, $E_x$ e $E_y$ não realizam qualquer trabalho virtual uma vez que, a elas não estão associados quaisquer deslocamentos virtuais.

- O momento $M_2$ ao estar aplicado a uma barra BC que, na sua configuração virtual deformada, não roda (apenas tem movimento de tranlação no plano), não produzirá qualquer trabalho virtual.

- Finalmente, as forças e os momentos que atuam no sistema real e que produzam  um trabalho virtual são apenas: $P$, $F_m$ e $M_1$.

**Princípio dos trabalhos virtuais**

Por aplicação do princípio dos trabalhos virtuais têm-se:

\begin{equation*}
\begin{aligned}
\delta U &= \delta U_{P} + \delta U_{F_m} + \delta U_{M_1} = 0\\
\Leftrightarrow~&
\vec{P}\cdot\vec{\delta}_B
+ \vec{M}_1\cdot\delta\vec{\theta}
- \vec{F}_m\cdot\vec{\delta}_D
= 0 \\
\Leftrightarrow~&
P\cdot\delta x_B
+ M_1\cdot\delta\theta
-F_m\cdot\delta x_D
= 0
\end{aligned}
\end{equation*}

Para desenvolver a expressão do trabalho virtual deve-se definir todos os deslocamentos virtuais que estão associados à força $P$ e à força $F_m$. Por outro lado, deve-se exprimir a força da mola em relação ao grau de liberdade do problema, a rotação $\theta$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P5/MAI_Ch09_P5b.jpg"
style="max-width: 100%;"/>

_Deslocamentos virtuais (pontos de aplicação das forças)_

_Relações geométricas_

\begin{equation*}
\begin{cases}
x_B = L\sin\theta
\quad\wedge\quad L= \overline{AB}=\overline{EC}\\
x_D = \displaystyle\frac{L}{2}\sin\theta
\end{cases}
\end{equation*}

Para obter-se a variação da posição do Ponto B e do Ponto D na direção das forças que neles atual, deve-se considera a derivada em relação a $\theta$:

\begin{equation*}
\begin{cases}
\delta x_B = L\cos\theta~\delta\theta \\
\delta x_D = \displaystyle\frac{L}{2}\cos\theta~\delta\theta
\end{cases}
\end{equation*}

_Força elástica da mola_

A força de uma mola é dada pela relação de proporcionalidade seguinte:

\begin{equation*}
F_m = k~\delta x_D = k~\displaystyle\frac{L}{2}~\sin\theta
\end{equation*}

Assim sendo, do PTV vem:

\begin{equation*}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
P\cdot\delta x_B
+ M_1\cdot\delta\theta
-F_m\cdot\delta x_D
= 0\\[1ex]
&~\Leftrightarrow~
P\cdot L\cos\theta\delta\theta
+ M_1\cdot\delta\theta
-(k\displaystyle\frac{L}{2}\sin\theta)
\cdot\displaystyle\frac{L}{2}\cos\theta\delta\theta = 0 \\
&~\Leftrightarrow~
P\cdot L \cos\theta
+ M_1
-k\frac{L^2}{4}\sin\theta\cos\theta = 0
\quad\wedge\quad \delta\theta \ne 0\\
&~\Leftrightarrow~
P\cdot L\cos\theta
+ M_1
-k\frac{L^2}{8}(2\sin\theta\cos\theta) = 0 \\
&~\Rightarrow~
\phi(\theta) = P\cdot L\cos\theta
+ M_1
-k\frac{L^2}{8}(\sin2\theta) = 0 \\
&~\Leftrightarrow~
\theta = 36.3^\circ
\end{aligned}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import nsolve, solve
import matplotlib.pylab as plt


# Define a switch for the color theme
use_light_theme = True

if use_light_theme:
    # Solarized Light Theme Colors
    background_color = (253/255, 246/255, 227/255)
    font_color = (0.1, 0.1, 0.1)
    grid_color = (147/255, 161/255, 161/255)
else:
    # Dark Theme Colors
    background_color = (.15, .15, .15)
    font_color = (.8, .8, .8)
    grid_color = (.2, .2, .2)

plt.rcParams['axes.facecolor'] = background_color
plt.rcParams['figure.facecolor'] = background_color
plt.rcParams['font.family'] = 'monospace'
plt.rcParams['font.size'] = 18
# plt.rcParams['text.usetex'] = True

params = {
    "ytick.color": font_color,
    "xtick.color": font_color,
    "grid.color": grid_color,
    "text.color": font_color,
    "axes.labelcolor": font_color,
    "axes.edgecolor": (.15, .15, .15)
}

plt.rcParams.update(params)

K = 60. # unit: N/m
BC = 0.4 # unit: m
DE = 0.2
P = 2 # unit: N
M1 = 0.5 # unit: N.m
M2 = 1.5 # unit: N.m
L = BC

t = sy.symbols('t')
EQ = P*L*sy.cos(t) + M1 - (K*L**2/8)*sy.sin(2*t)
sol = nsolve(sy.Eq(EQ,0),t,0.5)
theta = float(sol.num.real)
theta_deg = np.rad2deg(theta)
print(f' theta = {theta:.3f} radianos = {theta_deg:.1f} graus')

# alternativa:
# sol2 = solve(EQ,t) - esta solução conduz a 4 raizes sendo que a parte
# real de duas pelas são plausíveis

def ftheta(a): return P*L*np.cos(a) + M1 - (K*L**2/8)*np.sin(2*a)

vart1 = np.arange(-121,191,5)
vart2 = np.arange(0,91,5)
plt.figure(figsize=(9,6))
ax = plt.gca()
xx1, yy1 = vart1, ftheta(np.deg2rad(vart1))
plt.plot(xx1, yy1,'--',linewidth=4,color='brown',alpha=.5)
xx2, yy2 = vart2, ftheta(np.deg2rad(vart2))
plt.plot(xx2, yy2,'-',linewidth=4,color='brown',alpha=.9)
ax.axhline(y=0, color='gray')
ax.axvline(x=0, color='gray')
plt.plot(theta_deg, ftheta(theta),'o',markeredgecolor='r',markerfacecolor='g',
         markersize=22,alpha=0.6)
plt.xlabel(r"$\theta$ (graus)", size=18)
plt.ylabel(r"$\Phi(\theta)$ (N.m)", size=18)
plt.show()
```
