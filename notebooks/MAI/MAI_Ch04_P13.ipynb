- Condição de equilíbrio estático expressa no sumatório dos momentos das forças aplicadas em relação ao ponto A (de notar que as forças de reação nos eixos $x$, $y$ e $z$ não exergem momento em relação a A):

\begin{equation*}
\sum \vec{M}^R_A = \vec{0}
\qquad\therefore\qquad
\sum \vec{M}^R_A \cdot \vec{\lambda}_{AB} = 0
\end{equation*}

em que $\vec{\lambda_{AB}}$ representa o versor do eixo AB (e que contem o ponto A). De notar que o produto escalar deve ser zero na medida em que "a rotação em torno do eixo com a direção de AB é impedida pela tração do cabo CD".

Desenvolvendo a expressão do sumatório dos momentos em torno do ponto A:

\begin{equation*}
\sum \vec{M}^R_A = \vec{r}_{AE} \times \vec{F} + \vec{r}_{AC} \times \vec{T} + \vec{r}_{AB} \times \vec{B}
\qquad|\qquad \vec{r_{ij}} \equiv \vec{r}_{j/i} 
\end{equation*}

Contudo, o momento em torno do eixo com a direção de AB resulta,

\begin{equation*}
\sum \vec{M}^R_A \cdot \vec{\lambda}_{AB} = (\vec{r}_{AE} \times \vec{F} + \vec{r}_{AC} \times \vec{T} + \vec{r}_{AB} \times \vec{B}) \cdot \vec{\lambda}_{AB} = 0
\end{equation*}

desenvolvendo,

\begin{equation*}
(\vec{r}_{AE} \times \vec{F}) \cdot \vec{\lambda}_{AB} + (\vec{r}_{AC} \times \vec{T}) \cdot \vec{\lambda}_{AB} = 0
\quad\wedge\quad (\vec{r}_{AB} \times \vec{B}) \cdot \vec{\lambda}_{AB} = 
(\vec{\lambda}_{AB} \times \vec{r}_{AB} ) \cdot \vec{B} = 0 
\end{equation*}

Do enunciado é conhecido que,

\begin{equation*}
(\vec{r}_{AE} \times \vec{F}) \cdot \vec{\lambda_{AB}} = 10~\textrm{N.n}
\end{equation*}

Pelo que resulta,

\begin{equation*}
10 + (\vec{r}_{AC} \times \vec{T}) \cdot \vec{\lambda}_{AB} = 0
\end{equation*}

O vetor tração \vec{T} por sua vez pode ser obtido considerando o versor que define a sua diração, ou seja,

\begin{equation*}
\vec{T} = T \vec{\lambda}_{CD}
\end{equation*}

resultando a expressão,

\begin{equation*}
10  + (\vec{r}_{AC} \times T \vec{\lambda}_{CD}) \cdot \vec{\lambda}_{AB} = 0
\end{equation*}

onde,

\begin{equation*}
\vec{r}_{AC} = -x_C~\vec{i} + z_C~\vec{k}
\end{equation*}

\begin{equation*}
\vec{\lambda}_{CD} = \frac{\vec{CD}}{||\vec{CD}||} = \frac{\vec{r}_D - \vec{r}_C}{||\vec{r}_D - \vec{r}_C||}
= \frac{(x_D - x_C)~\vec{i} + (y_D - y_C)~\vec{j} + (z_D - z_C)~\vec{k}}{\sqrt{(x_D - x_C)²+ (y_D - y_C)² + (z_D - z_C)^2} }
\end{equation*}

\begin{equation*}
\vec{\lambda}_{AB} = \frac{\vec{AB}}{||\vec{AB}||} = \frac{\vec{r}_B}{||\vec{r}_B||}
= \frac{x_B~\vec{i} + y_B~\vec{j} + z_B~\vec{k}}{\sqrt{x_B²+y_B² + z_B^2} }
\end{equation*}

```python
import numpy as np

A = [0.,0.,0.]
B = [0,4.5,6.] # unit: m
C = [-3.,0,6.]
D = [-1.,2.5,0.]

def versor(a,b):
    v1 = np.subtract(b,a)
    modv1 = np.sqrt(np.dot(v1,v1))
    nv1 = v1/modv1
    return v1, modv1, nv1

vCD, modCD, nCD = versor(C,D)
print(f' vCD = {vCD} | norma de CD = {modCD:.3f} | versor dir. CD = {nCD} [m]')

vAB, modAB, nAB = versor(A,B)
print(f' vAB = {vAB} | norma de AB = {modAB:.3f} | versor dir. AB = {nAB} [m]')

rAC = np.subtract(C,A)
print(f' rAC = {rAC} [m]')
```

```python
import sympy as sy
from sympy.solvers import solve
from sympy.physics.mechanics import ReferenceFrame, cross, dot

S = ReferenceFrame('S') # ||x|| = ||y|| = ||z|| = 1
t = sy.symbols('t')

Mf = 10 # N.m

vecCD = nCD[0]*S.x + nCD[1]*S.y + nCD[2]*S.z
vecAB = nAB[0]*S.x + nAB[1]*S.y + nAB[2]*S.z
vecAC = rAC[0]*S.x + rAC[1]*S.y + rAC[2]*S.z
vT = t*vecCD

CP1 = cross(vecAC,vT); print('Cross Prod.: ',CP1)
SP1 = dot(CP1,vecAB); print('Dot Prod.: ',SP1)
expMA = SP1 + Mf
print('Exp MA = ',expMA)

sol = solve({expMA},{t})

T   = float(sol[t]);  print(f'     T = {T:.2f} [N]')
```
