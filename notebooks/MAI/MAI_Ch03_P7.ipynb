#### Problema 7

O cilindro representado na figura tem peso $\vec{P}$ e raio $r$, e o coeficiente de atrito estático em A e em B vale $\mu_\textrm{e}$. Determine a intensidade do maior binário $M$ que pode ser aplicado ao cilindro sem que este escorregue.

width="500">

#### Resolução

Comece-se por desenhar o diagrama de corpo livre (DCL) do cilindro:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P7/MAI_Ch03_P7a.jpg"
width="500">

A análise do equilíbrio estático do corpo permite escrever:

\begin{equation*}
\vec{R} = \sum_{i} \vec{F}_i = 0
\quad \wedge \quad
\vec{M}_{\textrm{C}}^{\textrm{R}}
= \sum_{i} \vec{M}_{\textrm{C}}
= \sum_{i} (\vec{r}_{\textrm{C}i} \times \vec{F}_i) = 0
\end{equation*}

em que o ponto $C$ representa o centroíde da secção circular onde se assume ser aplicada a força do peso próprio do objeto.

Desenvolvendo algebricamente as componentes cartesianas no plano das equações de equilíbrio estático resulta:

\begin{equation*}
\begin{cases}
\sum F_\textrm{x} &= 0 \\
\sum F_\textrm{y} &= 0 \\
\sum M_\textrm{C} &= 0 \\
\end{cases}~\Leftrightarrow~
\begin{cases}
N_\textrm{A} - F_\textrm{aB} &= 0 \\
F_\textrm{aA} + N_\textrm{B} - P &= 0 \\
M - F_\textrm{aA}\cdot r - F_\textrm{aB}\cdot r &= 0 \\
\end{cases}
\end{equation*}

com $F_\textrm{aA} = \mu_\textrm{e} N_\textrm{A}$ e $F_\textrm{aB} =
\mu_\textrm{e} N_\textrm{B}$, têm-se

\begin{equation*}
\begin{cases}
\sum F_\textrm{x} &= 0 \\
\sum F_\textrm{y} &= 0 \\
\sum M_\textrm{C} &= 0 \\
\end{cases}~\Leftrightarrow~
\begin{cases}
N_\textrm{A} - \mu_\textrm{e} N_\textrm{B} &= 0 \\
\mu_\textrm{e} N_\textrm{A} + N_\textrm{B} - P &= 0 \\
M - \mu_\textrm{e}(N_\textrm{A} + N_\textrm{B})r &= 0 \\
\end{cases}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import solve

NA, mue, NB, M, P, r = sy.symbols('NA mue NB M P r')

print('-------------------')
print('Sistema de equações')
print('-------------------')

sumFx = NA - mue*NB
print('sumFx = ',sumFx)
sumFy = mue*NA + NB - P
print('sumFy = ',sumFy)
sumMc = M - mue*(NA + NB)*r
print('sumMc = ',sumMc)

print('-------------------')
print('Solução')
print('-------------------')

sol = solve({sumFx,sumFy,sumMc},{M, NA, NB})
NA = sol[NA]; print('NA  = ',NA,' N')
NB = sol[NB]; print('NB  = ',NB,' N')
M = sol[M]; print('M   = ',M,' N.m')
```

Ou seja,

\begin{equation*}
\begin{cases}
N_\textrm{A} &=& \displaystyle\frac{\mu_\textrm{e}}{1+\mu_\textrm{e}^2}P\quad
\textrm{(N)}\\
N_\textrm{B} &=& \displaystyle\frac{1}{1+\mu_\textrm{e}^2}P\\
M &=& \displaystyle\frac{\mu_\textrm{e}(1+\mu_\textrm{e})}{1+\mu_\textrm{e}^2}\,
P\,r\quad\textrm{(N.m)}
\end{cases}
\end{equation*}
