Para o carregamento representado determine a força de tração em cada cabo e as reações em A (Desprezar o peso próprio da estrutura).

### Resolução

> Diagrama de corpo livre

Isole-se a estrutura soldada ao cursor A, pelo corte dos cabos CF e DE. Identifique-se no diagrama de corpo livre da estrutura as forças envolvidas (conhecidas e desconhecidas) nela diretamente aplicada:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch04/P1/MAI_Ch04_P1a.jpg"
width="500">

Identificam-se as incógnitas do problema:

1. no apoio $A$ : $R_{Ax}, R_{Az}$ e $M_{Ax}, M_{Az}$ (reações no apoio);

2. no apoio $D$ : $T_{DE}$ (força de tração no cabo DE);

3. no Ponto $C$ : $T_{CF}$ e $P$ (força de tração no cabo CF e força diretamente aplicada em C).

Desta forma, resulta o seguinte balanço em termos das componentes cartesianas das forças envolvidas no DCL:

* Forças incógnitas : $R_\textrm{Ax}, R_\textrm{Az}, M_\textrm{Ax}, M_\textrm{Az}, T_\textrm{CF}, T_\textrm{DE}$;

* Força conhecida : $P$.

Como do equilíbrio do corpo rígido conduz a 6 equações, convenientemente, existem 6 incógnitas.

> Equilíbrio estático

A condição de equilíbrio de um sólido rígido no espaço pressupõe:

\begin{equation*}
\bullet\quad\sum \vec{F} = \vec{0}~\Leftrightarrow~
\begin{cases}
\rightarrow(+)&x:&\sum F_x = 0 \\
\uparrow(+)&y:&\sum F_y = 0 \\
\swarrow(+)&z:&\sum F_z = 0 \\
\end{cases}
\end{equation*}

e

\begin{equation*}
\bullet\quad\sum \vec{M}_A = \vec{0}~\Leftrightarrow~
\begin{cases}
\vec{k}\times\vec{\jmath} = \vec{\imath}~(+)&:&\sum M_{Ax} = 0 \\
\vec{k}\times\vec{\imath} = \vec{\jmath}~(+)&:&\sum M_{Ay} = 0 \\
\vec{\imath}\times\vec{\jmath} = \vec{k}~(+)&:&\sum M_{Az} = 0 \\
\end{cases}
\end{equation*}

O ponto genérico A ($\sum \vec{M}_A$) que representa o polo sobre o qual é calculado o momento das forças externas, deve ser convenientemente escolhido.

Para estudar o equilíbrio estático da estrutura torna-se necessário escrever os vetores das forças, identificadas no diagrama de corpo livre, no referencial cartesiano escolhido:

\begin{equation*}
\begin{cases}
\vec{R}_\textrm{A} = R_\textrm{Ax}\,\vec{\imath} + 0\,\vec{\jmath}
+ R_\textrm{Az}\,\vec{z}  \\
\vec{T}_\textrm{CF} = T_\textrm{CF}\,\vec{\lambda}_\textrm{CF} \\
\vec{T}_\textrm{DE} = T_\textrm{DE}\,\vec{\lambda}_\textrm{DE} \\
\vec{F} = -P\,\vec{\jmath} = -480\,\vec{\jmath} \\
\vec{M}_\textrm{Ax} = M_\textrm{Ax}\,\vec{\imath} \\
\vec{M}_\textrm{Az} = M_\textrm{Az}\,\vec{k} \\
\end{cases}
\end{equation*}

os versores $\vec{\lambda}$ definem a direção e sentido onde atuam as forças de tração nos cabos CF e DE. As componentes cartesianas de $\vec{\lambda}$ são os cossenos diretores, definidos à custa dos ângulos formados entre a direção da força e os eixos coordenados ($\vec{\imath},\vec{\jmath},\vec{k}$). É possível determinar as componentes cartesiandas dos versores $\vec{\lambda}$ usando coordenadas de dois pontos contidos na linha acção das forças: ponto de aplicação e o ponto de extremidade do segmento de reta orientado. Trabalhando sobre os dados do problema vem:

> Pontos

\begin{equation*}
\begin{aligned}
&C(80,0,135)~;~
D(80,0,90)\\
&E(80,120,0)~;~
F(0,60,135)
\end{aligned}
\end{equation*}

```python
import numpy as np

A = [0,0,0]
C = [80,0,135]
D = [80,0,90]
E = [80,120,0]
F = [0,60,135]

P = 480 # unit:N
```

> Vetores - componentes cartesianos e módulos:

* $\vec{CF}$:

\begin{equation*}
\vec{CF} = \vec{OF} - \vec{OC} = (0,60,135) - (80,0,135)
= -80\,\vec{\imath} + 60\,\vec{\jmath} + 0\,\vec{k}\\
\end{equation*}

```python
vCF = np.subtract(F,C); print('vCF = ',vCF)
```


\begin{equation*}
||\vec{CF}|| = \sqrt{CF_x^2+CF_y^2+CF_z^2}
= \sqrt{(-80)^2 + 60^2} = 100.0~\textrm{mm}
\end{equation*}

```python
modCF = np.sqrt(np.dot(vCF,vCF))
print('modulo de CF = ','{:.1f}'.format(modCF),' mm')
```

* $\vec{DE}$:

\begin{equation*}
\vec{DE} = \vec{OE} - \vec{OD} = (80,120,0) - (80,0,90)
= 0\,\vec{\imath} + 120\,\vec{\jmath} +(-90)\,\vec{k}\\
\end{equation*}

```python
vDE = np.subtract(E,D); print('vDE = ',vDE)
```

\begin{equation*}
|| \vec{DE}|| = \sqrt{DE_x^2+DE_y^2+DE_z^2}
= \sqrt{(120)^2 + (-90)^2} = 150.0~\textrm{mm}\\
\end{equation*}

```python
modDE = np.sqrt(np.dot(vDE,vDE))
print('modulo de DE = ','{:.1f}'.format(modDE),' mm')
```

Assim sendo os vetores das forças aplicadas em C vêm:

\begin{equation*}
\begin{aligned}
\bullet~\vec{T}_\textrm{CF} = T_\textrm{CF}\,\vec{\lambda}_\textrm{CF}
= T_\textrm{CF}\frac{\vec{CF}}{||\vec{CF}||} =
T_\textrm{CF}(-0.8\,\vec{\imath} + 0.6\,\vec{\jmath} + 0\,\vec{k})
\end{aligned}
\end{equation*}

```python
np.set_printoptions(precision=3)
verCF = vCF/modCF
print('verCF = ',verCF)
print('Verifications: ----------')
print('norma verCF:',np.sqrt(np.dot(verCF,verCF)))
```

\begin{equation*}
\begin{aligned}
\bullet~\vec{T}_\textrm{DE} = T_\textrm{DE}\,\vec{\lambda}_\textrm{DE}
= T_\textrm{DE}\frac{\vec{DE}}{||\vec{DE}||} =
T_\textrm{DE}(0\,\vec{\imath} + 0.8\,\vec{\jmath} - 0.6\,\vec{k})
\end{aligned}
\end{equation*}

```python
np.set_printoptions(precision=3)
verDE = vDE/modDE
print('verDE = ',verDE)
print('Verifications: ----------')
print('norma verDE:',np.sqrt(np.dot(verDE,verDE)))
```

> Desenvolvendo as equações escalares do equilíbrio das forças

\begin{equation*}
\bullet
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0 \\
\sum F_z = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
R_\textrm{Ax} -0.8\,T_\textrm{CF} = 0 \\
0.6\,T_\textrm{CF} + 0.8\,T_\textrm{DE} - P = 0 \\
R_\textrm{Az} -0.6\,T_\textrm{DE} = 0 \\
\end{cases}
\end{equation*}

> Desenvolvendo as equações escalares do equilíbrio dos momentos das forças

Do equilíbrio estático sabe-se que, se um corpo for submetido a um sistema de forças, o momento resultante dessas forças em relação a um ponto O, pode ser
determinado pela adição dos vetores momento de cada força. Este resultado pode ser escrito sumariamente por:

\begin{equation*}
\bullet\quad \left(\vec{M}_R\right)_A \,=\,
\left(\sum_i \vec{M_{F_i}}\right)_A
= \sum_i \vec{r}_{Ai}\times\vec{F_i} = \vec{0}
\end{equation*}

Desenvolvendo para o caso em estudo,

\begin{equation*}
\begin{aligned}
&\sum_{i=1}^3 \vec{r}_{Ai}\times\vec{F_i} +
\left(\vec{M_{A_j}} \right)_{\{j=x,z\}} = \vec{0} \\
~\Leftrightarrow~&
(\vec{r}_{\textrm{AC}}\times\vec{P}) +
(\vec{r}_{\textrm{AC}}\times\vec{T}_\textrm{CF}) +
(\vec{r}_{\textrm{AD}}\times\vec{T}_\textrm{DE}) +
\vec{M}_\textrm{Ax} + \vec{M}_\textrm{Az} = \vec{0}
\end{aligned}
\end{equation*}

Os vetores posição entre o polo A e os pontos de aplicação das forças que produzem momento em relação a esse polo têm as seguintes componentes
cartesianas:

\begin{equation*}
\begin{aligned}
\vec{r}_{\textrm{AC}} = \vec{OC} - \vec{OA} = 80&\vec{\imath}~+~0\vec{\jmath} +~135\,
\vec{k} \\
\vec{r}_{\textrm{AD}} = \vec{OD} - \vec{OA} = 80&\vec{\imath}~+~0\vec{\jmath} +~90\,
\vec{k} \\
\end{aligned}
\end{equation*}

```python
rAC =  np.subtract(C,A); print('rAG = ',rAC)
rAD =  np.subtract(D,A); print('rAC = ',rAD)
```

O cálculo do produto vetorial entre vetores pode ser calculado pelo
determinante de uma matriz de acordo com o seguinte formulário:

\begin{equation*}
(\vec{r}_{\textrm{AC}}\times\vec{P}) +
(\vec{r}_{\textrm{AC}}\times\vec{T}_\textrm{CF}) +
(\vec{r}_{\textrm{AD}}\times\vec{T}_\textrm{DE}) +
\vec{M}_\textrm{Ax} + \vec{M}_\textrm{Az} = \vec{0}
\end{equation*}

\begin{equation*}
\bullet\quad\vec{r}_{\textrm{AC}}\times\vec{P} =
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
x_{\textrm{AC}} & 0 & z_{\textrm{AC}} \\
0 & -P & 0
\end{vmatrix} = 64800\vec{\imath} + 0\vec{\jmath} + (- 38400)\vec{k}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import solve
from sympy.physics.mechanics import ReferenceFrame, cross

S = ReferenceFrame('S') # ||x|| = ||y|| = ||z|| = 1
vAC = rAC[0]*S.x + rAC[1]*S.y + rAC[2]*S.z
vP = -P*S.y
CP1 = cross(vAC,vP); print('Cross 1: ',CP1)
```

\begin{equation*}
\bullet\quad\vec{r}_{\textrm{AC}}\times\vec{T}_\textrm{CF} =
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
x_{\textrm{AC}} & 0 & z_{\textrm{AC}} \\
-0.8T_{\textrm{CF}} & 0.6T_{\textrm{CF}} & 0
\end{vmatrix} =
T_{\textrm{CF}}(-81\vec{\imath} - 108\vec{\jmath} + 48\vec{k})
\end{equation*}

```python
tcf = sy.symbols('tcf')
vCF = tcf*(verCF[0]*S.x + verCF[1]*S.y + verCF[2]*S.z)
CP2 = cross(vAC,vCF); print('Cross 2: ',CP2)
```

\begin{equation*}
\bullet\quad\vec{r}_{\textrm{AD}}\times\vec{T}_\textrm{DE} =
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
x_{\textrm{DE}} & 0 & z_{\textrm{DE}} \\
0 & 0.8T_{\textrm{DE}} & -0.6T_{\textrm{DE}}
\end{vmatrix} =
T_{\textrm{DE}}(-72\vec{\imath} + 48\vec{\jmath} + 64\vec{k})
\end{equation*}

```python
tde = sy.symbols('tde')
vAD = rAD[0]*S.x + rAD[1]*S.y + rAD[2]*S.z
vDE = tde*(verDE[0]*S.x + verDE[1]*S.y + verDE[2]*S.z)
CP3 = cross(vAD,vDE); print('Cross 3: ',CP3)
```

\begin{equation*}
\bullet\quad\sum \vec{M}_A = \sum (\vec{r}\times\vec{F}) = \vec{0}
\end{equation*}

\begin{equation*}
\begin{aligned}
~\Leftrightarrow~ &(64800 - 81\,T_\textrm{CF} - 72\,T_\textrm{DE} +
M_\textrm{Ax})\,
\vec{\imath} \\
&+ (-108\,T_\textrm{CF} + 48\,T_\textrm{DE})\,\vec{\jmath} \\
& + (-38400 + 48\,T_\textrm{CF} + 64\,T_\textrm{DE} + M_\textrm{Az})\,\vec{k}
= \vec{0}
\end{aligned}
\end{equation*}

```python
maxx, maz = sy.symbols('maxx maz')
MAX = maxx*S.x
MAZ = maz*S.z
expMA = CP1 + CP2 + CP3 + MAX + MAZ
print('Exp MA = ',expMA)
```

Decompondo estas quantidades vetoriais nas equações escalares do equilíbrio dos momentos produzidos pelas forças externas aplicadas resulta o seguinte sistema de equações:

\begin{equation*}
\bullet\quad\sum \vec{M}_A = \vec{0}~\Leftrightarrow~
\begin{cases}
\sum M_{Ax} = 0 \\
\sum M_{Ay} = 0 \\
\sum M_{Az} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
64800 - 81\,T_\textrm{CF} - 72\,T_\textrm{DE} +
M_\textrm{Ax}= 0 \\
-108\,T_\textrm{CF} + 48\,T_\textrm{DE} = 0 \\
-38400 + 48\,T_\textrm{CF} + 64\,T_\textrm{DE} + M_\textrm{Az} = 0
\end{cases}
\end{equation*}

Resulta finalmente o seguinte sistema de equações (6 equações a 6 incógnitas):

\begin{equation*}
\bullet\quad
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0 \\
\sum F_z = 0 \\
\sum M_{Ax} = 0 \\
\sum M_{Ay} = 0 \\
\sum M_{Az} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
R_\textrm{Ax} -0.8\,T_\textrm{CF} = 0 \\
0.6\,T_\textrm{CF} + 0.8\,T_\textrm{DE} - P = 0 \\
R_\textrm{Az} -0.6\,T_\textrm{DE} = 0 \\
64800 - 81\,T_\textrm{CF} - 72\,T_\textrm{DE} +
M_\textrm{Ax}= 0 \\
-108\,T_\textrm{CF} + 48\,T_\textrm{DE} = 0 \\
-38400 + 48\,T_\textrm{CF} + 64\,T_\textrm{DE} + M_\textrm{Az} = 0
\end{cases}
\end{equation*}

Este sistema de equações pode ser resolvido pela algebra de sistema:

```python
rax, raz = sy.symbols('rax raz')

EFX = rax + tcf*verCF[0]
EFY = tcf*verCF[1] + tde*verDE[1] - P
EFZ = raz + tde*verDE[2]
EMX = expMA.args[0][0].args[2][0]
EMY = expMA.args[0][0].args[2][1]
EMZ = expMA.args[0][0].args[2][2]

sol = solve({EFX,EFY,EFZ,EMX,EMY,EMZ},{rax,raz,maxx,maz,tcf,tde})

RAx = float(sol[rax])
print('RAx = ','{:.1f}'.format(RAx),' N')
RAz = float(sol[raz])
print('RAz = ','{:.1f}'.format(RAz),' N')
MAx = float(sol[maxx])
print('MAx = ','{:.1f}'.format(MAx),' N.mm | ', '{:.1f}'.format(MAx/1e3),' N.m')
MAz = float(sol[maz])
print('MAz = ','{:.1f}'.format(MAz),' N.mm')
TCF = float(sol[tcf])
print('TCF = ','{:.1f}'.format(TCF),' N')
TDE = float(sol[tde])
print('TDE = ','{:.1f}'.format(TDE),' N')
```

O sistema de equações anterior pode também ser resolvido usando algebra
matricial. Comece-se por reescrever o sistema na seguinte forma matricial:

\begin{equation*}
[A]\{x\} = \{b\}
\Leftrightarrow
\begin{bmatrix}
% RAx RAz MAx MAz TCF TDE
1 & 0 & 0 & 0 & -0.8 & 0 \\
0 & 0 & 0 & 0 & 0.6 & 0.8 \\
0 & 1 & 0 & 0 & 0 & -0.6 \\
0 & 0 & 1 & 0 & -81 & -72 \\
0 & 0 & 0 & 0 & -108 & 48 \\
0 & 0 & 0 & 1 & 48 & 64
\end{bmatrix}
\begin{Bmatrix}
R_\textrm{Ax} \\
R_\textrm{Az} \\
M_\textrm{Ax} \\
M_\textrm{Az} \\
T_\textrm{CF} \\
T_\textrm{DE}
\end{Bmatrix} =
\begin{Bmatrix}
0 \\
480 \\
0 \\
-64800 \\
0 \\
38400
\end{Bmatrix}
\end{equation*}

Construção da matriz de coeficientes $[A]$:

```python
np.set_printoptions(precision=3, suppress=True)

A1 = [1,0,0,0,-0.8,0]
A2 = [0,0,0,0,0.6,0.8]
A3 = [0,1,0,0,0,-0.6]
A4 = [0,0,1,0,-81,-72]
A5 = [0,0,0,0,-108,48]
A6 = [0,0,0,1,48,64]

A = np.array([A1,A2,A3,A4,A5,A6])
print(f'A =\n {A}')
```

Construção da vetor $\{b\}$:

```python
b = np.array([0,P,0,-64600,0,38400])
print(f'b = {b}')
```

A solução do problema passa por calcular o seguinte produto:

\begin{equation*}
\{x\} \equiv
\begin{Bmatrix}
R_\textrm{Ax} \\
R_\textrm{Az} \\
M_\textrm{Ax} \\
M_\textrm{Az} \\
T_\textrm{CF} \\
T_\textrm{DE}
\end{Bmatrix}
= [A]^{-1}\{b\}
=
\begin{bmatrix}
% RAx RAz MAx MAz TCF TDE
1 &  0.333 & 0 & 0 & -0.006 &  0 \\
0 &  0.562 & 1 & 0 & 0.003  & 0 \\
0 &  101.25 & 0 & 1 & -0.187 & 0 \\
0 &  -80.0 & 0 & 0 & 0 & 1 \\
0 &  0.417 & 0 & 0 & -0.007 & 0 \\
0 &  0.938 & 0 & 0 & 0.005 & 0
\end{bmatrix}
\begin{bmatrix}
0 \\
480 \\
0 \\
-64800 \\
0 \\
38400
\end{bmatrix}
\end{equation*}

```python
invA = np.linalg.inv(A)
print(f'A^-1 =\n {invA}')

print('\ncode 1 --------------')
x = np.linalg.solve(A,b)
print(f'x = {x} [N / N.mm]')

print('\ncode 2 --------------')
x2 = np.dot(invA,b)
print(f'x = {x2} [N / N.mm]')
```

### Apêndice

Qualquer sistema de forças, $F_1, F_2, \ldots, F_n$ pode ser substituído por um sistema força-binário equivalente aplicado num dado ponto $O$, constituído pelos vetores $\vec{R}$ e $M_0^R$. Quando as forças que actuam sobre um corpo podem ser reduzidas, num ponto arbitrário $O$, a um sistema força-binário constituído por uma força nula e um binário nulo, diz-se que o corpo está em equilíbrio estático.

As condições necessárias e suficientes para garantir o equilíbrio são:

\begin{equation*}
\begin{cases}
\sum\vec{F} = \vec{0}\\
\sum\vec{M}_0 = \sum\vec{r}\times\vec{F} = \vec{0}
\end{cases}
\end{equation*}

Decompondo a força e o momento nas suas componentes cartesianas, estas duas equações vetoriais conduzem às seguintes seis equações escalares:

\begin{equation*}
\begin{cases}
~\sum F_x = 0\\
~\sum F_y = 0\\
~\sum F_z = 0\\
\sum M_x = 0 \\
\sum M_y = 0 \\
\sum M_z = 0
\end{cases}
\end{equation*}

Para escrever as equações de equilíbrio de um corpo rígido é essencial identificar todas as forças aplicadas no corpo.  Essas forças devem constar do
diagrama de corpo livre.
