#### Problema 8

A janela representada na figura tem a massa de 30 kg e está ligada a duas dobradiças situadas nos cantos A e B. O telhado forma com o plano horizontal um ângulo de 30$^\circ$, e a janela é mantida na posição horizontal indicada por uma barra EC. Considere que a força exercida pela barra sobre a janela tem a
direção EC e que a dobradiça em A não exerce qualquer força na direção do seu eixo.

a) Classifique o problema como hipoestático, isostático ou hiperestático? Justifique.

b) Determine a força exercida pela barra EC e as reações nas dobradiças A e B.
