#### Resolução

Considere-se o deslocamento virtual da estrutura representado na figura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Au13/A13_P5a.jpg"
width="500" height="350">

Por aplicação do princípio dos trabalhos virtuais:

\begin{equation}
\begin{aligned}
\delta U &= \delta U_{F_S} + \delta U_{M_1}
+ \delta U_{P} = 0\\
\Leftrightarrow~&
\vec{F}_S\cdot\delta\vec{x}_D +
\vec{M}_1\cdot\delta\vec{\theta} +
\vec{P}\cdot\delta\vec{x}_B = 0 \\
\Leftrightarrow~&
-F_S\cdot\delta x_D +
M_1\cdot\delta\theta +
P\cdot\delta x_B = 0
\end{aligned}
\end{equation}

De notar que a barra BC onde atua o momento $M_2$ não roda no deslocamento
virtual considerado, apenas tem movimento de translação no plano, pelo que o seu
trabalho virtual é nulo.

_Deslocamentos virtuais (pontos de aplicação das forças)_

_Relações geométricas_

\begin{equation}
\begin{cases}
x_B = \overline{BC}\sin\theta \\[2ex]
x_D = \overline{DE}\sin\theta
\end{cases}
\end{equation}

Por derivação em ordem ao parâmetro $\theta$:

\begin{equation}
\begin{cases}
\delta x_B = \overline{BC}\cos\theta\delta\theta \\[2ex]
\delta x_D = \overline{DE}\cos\theta\delta\theta
\end{cases}
\end{equation}

_Força elástica da mola_

A força de uma mola é dada pela relação de proporcionalidade seguinte:

\begin{equation}
F_S = k \,\delta x_D = k\,\overline{DE}\sin\theta
\end{equation}

- PTV:

\begin{equation}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
-F_S\cdot\delta x_D +
M_1\cdot\delta\theta +
P\cdot\delta x_B = 0\\
&~\Leftrightarrow~
-k\,\overline{DE}\sin\theta\cdot\overline{DE}\cos\theta\delta\theta
+ M_1\cdot\delta\theta
+ P\cdot\overline{BC}\cos\theta\delta\theta = 0 \\
&~\Leftrightarrow~
-k\,\overline{DE}^2\sin\theta\cos\theta
+ M_1
+ P\cdot\overline{BC}\cos\theta = 0\\
&~\Leftrightarrow~
-\frac{k\,\overline{DE}^2}{2}(2\sin\theta\cos\theta)
+ M_1
+ P\cdot\overline{BC}\cos\theta = 0 \\
&~\Leftrightarrow~
-\frac{k\,\overline{DE}^2}{2}(\sin2\theta)
+ M_1
+ P\cdot\overline{BC}\cos\theta = 0 \\
&~\Leftrightarrow~
\theta = 36.3^\circ
\end{aligned}
\end{equation}

```python
import numpy as np
import sympy as sy
from sympy.solvers import nsolve

K = 60. # unit: N/m
BC = 0.4 # unit: m
DE = 0.2
P = 2 # unit: N
M1 = 0.5 # unit: N.m
M2 = 1.5 # unit: N.m

t = sy.symbols('t')

EQ = -(K*DE**2/2)*sy.sin(2*t) + M1 + P*BC*sy.cos(t)
sol = nsolve(sy.Eq(EQ,0),t,0.5)
theta = float(sol.num.real)
print(' theta = ','{:.3f}'.format(theta), 'radianos')
print('       = ','{:.1f}'.format(np.rad2deg(theta)), 'graus')
```
