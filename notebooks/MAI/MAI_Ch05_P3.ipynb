### Resolução

**Solução integral**

Comece-se por notar que o centroide da região está localizado segundo o eixo dos $y$, pois este é um eixo de simetria. Matematicamente, esta condição expressa-se por:

\begin{equation*}
\overline{x} = \frac{Q_y}{A}
= \frac{\displaystyle\int x\,\mathrm{d}A}{\displaystyle\int\mathrm{d}A} = 0
\end{equation*}

Conduto, para uma verificação por solução integral, note-se que, como ilustrado na figura abaixo, para um elemento infinitesimal $\mathrm{d}A$ a uma distância $x$ do eixo $y$, se consegue encontrar um outro elemento $\mathrm{d}A$ igual à distância $–x$ do eixo $y$. Consequentemente o integral será zero.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P3/MAI_Ch05_P3a.jpg"
width="700">

Como se pode confirmar:

\begin{equation*}
\begin{aligned}
Q_y &= \int_A x\,\mathrm{d}A
= \int_r\int_\theta x\,r\,\mathrm{d}\theta\mathrm{d}r\Big|_{\large x = r\cos\theta}
= \int_r\int_\theta r^2\cos\theta\,\mathrm{d}\theta\mathrm{d}r \\
& = \int_{r_1}^{r_2} r^2 \left(\int_{0}^{\pi} \cos\theta\,
\mathrm{d}\theta\right)\mathrm{d}r
 = \int_{r_1}^{r_2} r^2 [\sin\theta]_{0}^{\pi} \,
\mathrm{d}r\\
& = \int_{r_1}^{r_2} r^2 [\sin\pi - (\sin 0)] \,
\mathrm{d}r
= 0
\end{aligned}
\end{equation*}

logo,

\begin{equation*}
\overline{x} = \frac{Q_y}{A} = 0 ~\textrm{mm}
\end{equation*}

Para a resolução da equação integral no cálculo da ordenada $\overline{y}$:

\begin{equation*}
\overline{y} = \frac{Q_x}{A}
= \displaystyle\frac{\displaystyle\int y\,\mathrm{d}A}{\displaystyle\int\mathrm{d}A}
\end{equation*}

comece-se por considerar o elemento de área $\mathrm{d}A = r\mathrm{d}\theta\mathrm{d}r$ (em coordenadas polares, mais adequadas para a geometria do problema proposto) representado na figura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P3/MAI_Ch05_P3b.jpg"
width="700">

A área da superfície é calculado desenvolvendo algebricamente a seguinte expressão:

\begin{equation*}
\begin{aligned}
A &= \int\mathrm{d}A =
\int_{r_1}^{r_2}\int_{0}^{\pi} r\,\mathrm{d}\theta\mathrm{d}r
= \int_{r_1}^{r_2}r\left(\int_{0}^{\pi} \,\mathrm{d}\theta\right)\mathrm{d}r \\
&= \int_{r_1}^{r_2}r [\theta]_{0}^{\pi} \mathrm{d}r
= \pi\int_{r_1}^{r_2}r \mathrm{d}r
= \pi\left[\frac{r^2}{2}\right]_{r_1}^{r_2}
= \frac{\pi}{2}\left(r_2^2 - r_1^2\right)
\end{aligned}
\end{equation*}

O momento estático ou de primeira ordem em relação ao eixo $x$ vem:

\begin{equation*}
\begin{aligned}
Q_x &= \int y\,\mathrm{d}A
= \int\int y\,r\,\mathrm{d}\theta\mathrm{d}r\Big|_{\large y = r\sin\theta}
= \int\int r^2\sin\theta\,\mathrm{d}\theta\mathrm{d}r \\
& = \int_{r_1}^{r_2} r^2 \left(\int_{0}^{\pi} \sin\theta\,
\mathrm{d}\theta\right)\mathrm{d}r
 = \int_{r_1}^{r_2} r^2 [-\cos\theta]_{0}^{\pi} \,
\mathrm{d}r\\
& = \int_{r_1}^{r_2} r^2 [-\cos\pi - (-\cos 0)] \,
\mathrm{d}r = 2 \left[\frac{r^3}{3}\right]_{\small r_1}^{\small r_2}
= \frac{2}{3} \left(r_2^3 - r_1^3\right)
\end{aligned}
\end{equation*}

Pelo que,

\begin{equation*}
\overline{y} = \frac{Q_x}{A} =
\frac{4}{3\pi}\left(\frac{r_2^3 - r_1^3}{r_2^2 - r_1^2}\right)
= 6.451 ~\textrm{mm}
\end{equation*}

**Solução direta**

Considere-se a seguinte estratégia de resolução que passa pela subtração de duas semi-esferas que diâmetro $r_1$ e $r_2$ para a obtenção da superfície final:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P3/MAI_Ch05_P3c.jpg"
width="900">

O centroide e área de cada semi-esfera pode ser obtido diretamente de tabelas em
função do raio $r$:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P3/MAI_Ch05_P3d.jpg"
width="900">

| Sub-áreas|$A_i$, mm$^2$|$\overline{x}_i$, mm|$\overline{x}_iA_i$, mm$^3$|
|:-------:|:----------:|:-----------------:|:------------------------:|
|(1)|$\displaystyle\frac{\pi r_1^2}{2}$|$\displaystyle\frac{4r_1}{3\pi}$|$\displaystyle\frac{2r_1^3}{3}$|
|(2)|$\displaystyle\frac{\pi r_2^2}{2}$|$\displaystyle\frac{4r_2}{3\pi}$|$\displaystyle\frac{2r_2^3}{3}$|
|$\sum: (2) - (1)$|$\displaystyle\frac{\pi}{4}(r_2^2 - r_1^2)$|-|$\displaystyle\frac{2}{3}(r_2^3 - r_1^3)$|

Sendo que,

\begin{equation*}
\overline{y} = \frac{Q_x}{A}
= \frac{\int x\,\mathrm{d}A}{\int\mathrm{d}A}
= \frac{\frac{1}{3}(r_2^3 - r_1^3)}{\frac{\pi}{4}(r_2^2 - r_1^2)}
= \frac{4}{3\pi}\left(\frac{r_2^3 - r_1^3}{r_2^2 - r_1^2}\right)
= 6.451 ~\textrm{mm}
\end{equation*}

```python
import numpy as np
import sympy as sy

r1v, r2v = 8, 12 # unit: mm
r1, r2 = sy.symbols('r1 r2')

A1 = np.pi*r1**2/2
print('A1 = ',A1,' m2')
A2 = np.pi*r2**2/2
print('A2 = ',A2,' m2')
x1 = 4*r1/3/np.pi
print('x1 = ',x1,' m')
x2 = 4*r2/3/np.pi
print('x2 = ',x2,' m')

A = A2 - A1
print('A = ',sy.simplify(A),' m2')
Qx = A2*x2 - A1*x1
print('Qx =',Qx,' m3')

xcoor1 = Qx/A
print('xcoor (tabelas) =',sy.simplify(xcoor1),' m')
xcoor1v1 = xcoor1.subs([(r1,r1v),(r2,r2v)])
print('xcoor (tabelas) =','{:.3f}'.format(xcoor1v1),' mm')
```
