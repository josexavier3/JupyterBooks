#### a) Trata-se de uma treliça simples?

Para a treliça ser simples é necessário que seja constituída por triângulos (rígidos) e ser globalmente isostática. Conforme se pode ver pela figura abaixo, os triângulos a amarelo (EOML e ELKP) não são rígidos porque possuem mais de três nós, que neste caso, são os nós em O e P. Por conseguinte, a treliça não é simples.

#### b) Classifique a treliça globalmente.

Para determinar a estatia global da treliça (g), recorre-se à expressão seguinte:

\begin{equation*}
g = 2n - (m+r)
\end{equation*}

onde,

- $n$ é o número de nós barras da treliça;
- $m$ é o número de barras;
- $r$ é o número de reacções ou ligações ao exterior .


```python
n, m, r = 16, 29, 3
print(f'2n = {2*n:.0f} | m  = {m:.0f} | r  = {r:.0f}')

def grauhiperstaticidade(nn,mm,rr):
    return 2*nn - (mm+rr)

g = grauhiperstaticidade(n,m,r)
print(f'g = {g:.0f}')
```

Pelo que, a treliça é globalmente isostática.

width="500">


**Estatia global**

Estatia global da treliça ($\alpha_g$) é calculada de acordo com:

\begin{equation*}
\alpha_g = \alpha_e + \alpha_i
\end{equation*}

em que $\alpha_e$ e $\alpha_i$ são a estatia exterior e estatia interior da treliça calculadas, respectivamente, por:

\begin{equation*}
\alpha_e = 3 × \textrm{( nº de apoios − 1)} − \textrm{nº de libertações exteriores}
\end{equation*}

- Apoios: O número de apoios refere-se ao total de apoios que a treliça tem
- Libertações Exteriores: Este termo refere-se ao número de graus de liberdade que não estão restringidos pelos apoios

\begin{equation*}
\alpha_i = 3 × \textrm{nº de malhas fechadas} − \textrm{nº de ligações interiores}
\end{equation*}

- Malha fechada numa treliça refere-se a uma região da estrutura que é completamente delimitada por barras, formando um circuito fechado. Malhas fechadas considera que cada malha adiciona três graus de liberdade internos.
- Ligações interiores referem-se às barras internas da treliça que conectam os nós entre si.

```python
def estatiaexterior(napoios, nlibertacoes):
    return 3*(napoios-1) - nlibertacoes

def estatiainterior(nmalhas,nligainter):
    return  3*nmalhas - nligainter

npoios_, nlibext_ = 2, 3
alphe = estatiaexterior(npoios_,nlibext_)
print(f'estatia exterior = 3({npoios_} -1) - {nlibext_} =  {alphe}')
nmalh_, nlibint_ = 14, 1*2 + 2*5 + 3*6 + 4*3 
alphi = estatiainterior(nmalh_,nlibint_)
print(f'estatia interior = 3({nmalh_}) - {nlibint_} = {alphi}')
estatiaglobal = alphe + alphi
print((f'estatia global = {estatiaglobal}'))
```

Para calcular as reacções da treliça, começa-se por escrever as equações de equilíbrio estático relativas ao somatório de forças em $x$ e $y$ assim como a expressão do somatório dos momnetos em relação ao ponto I.

\begin{equation*}
\begin{cases}
\rightarrow(+)&:&\sum F_x = 0 \\
\uparrow(+)&:&\sum F_y = 0 \\
\circlearrowleft(+)&:&\sum M_A =0
\end{cases}~\Leftrightarrow~
\begin{cases}
-F_E \sin\theta - F_F\sin\theta - F_G\sin\theta + R_{Ix} = 0  \\
-F_E \cos\theta - F_F \cos\theta - F_G \cos\theta + R_{Iy} + R_{Ay} = 0  \\
- R_{Ay} (6L) + F_E (3L/\cos\theta) + F_F(3L\cos\theta) + F_G(2L\cos\theta) = 0
\end{cases}
\end{equation*}


```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

def printsymSF(expr):
    expr2 = sy.simplify(sy.factor(expr))
    return expr2.xreplace({n : round(n, 3) for n in expr.atoms(sy.Number)})

def c(a): return np.cos(a)
def s(a): return np.sin(a)

L, a = 5., np.deg2rad(30.) # unit: m, rad
FE, FF, FG = 10., 20., 10. # unit kN

rix, riy, ray = sy.symbols('rix riy ray')

SFX = -FE*s(a) - FF*s(a) - FG*s(a) + rix
print('sum Fx :', printsymSF(SFX))
SFY = -FE*c(a) - FF*c(a) - FG*c(a) + riy + ray
print('sum Fy :', printsymSF(SFY))
SMI = - ray*6*L + FE*(3*L/c(a)) + FF*(3*L*c(a)) + FG*(2*L*c(a))
print('sum MI :', printsymSF(SMI))
```

```python
sol = solve({SFX,SFY,SMI},{ray, rix, riy})

RAy = sol[ray]
print(f'RAy    = {RAy:.1f} kN')
RIx = sol[rix]
print(f'RIx    = {RIx:.1f} kN')
RIy = sol[riy]
print(f'RIy    = {RIy:.1f} kN')
```

#### d) Diga qual a vantagem do método das secções em relação ao método dos nós.

O método dos nós requer que se analise o equílibrio dos vários nós antes de se determinarem as forças nas barras mais centrais da treliça. Já no método das secções, ao dividir-se a treliça em duas partes através de um corte imaginário, como as barras estão sujeitas apenas a tracção ou compressão (esforços normais), estes podem ser calculados a partir das condições de equilíbrio estático de uma das partes seccionada da treliça, tornando assim, o método das secções mais simples de aplicar quando se pretende calcular as forças instaladas em algumas barras da treliça, sejam elas barras mais centrais da treliça ou não.



#### e) Utilize o método das secções para determinar a força a que estão sujeitas as barras ML e CD, indicando também se estão à tração ou à compressão.

Recorrendo ao método das secções para calcular as forças que as barras ML e CD estão sujeitas secciona-se a treliça de acordo com a ilustração da figura abaixo.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch07/P4/MAI_Ch07_P4f.png"
width="700">

Escreve-se a equação dos momentos relativamente ao ponto E; e escreve-se também a equação dos momentos relativamente ao ponto M:

\begin{equation*}
\begin{cases}
\circlearrowleft(+)&:&\sum M_E =0\\
\circlearrowleft(+)&:&\sum M_M =0\\
\rightarrow(+)&:&\sum F_x = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
- R_{Ay} (3L) + N_{ML} (3L\tan\theta) = 0 \\
- R_{Ay} (2L) + N_{DE} (2L\sin\theta) = 0 \\
N_{DE}\cos\theta + N_{OE}\cos2\theta + N_{ML} = 0
\end{cases}
\end{equation*}

Note-se que, como os nós C, D e E são colineares e não havendo qualquer força aplicada ao nó D, a força instalada na barra CD é a mesma que está instalada na força na barra DE, pelo que:


\begin{equation*}
N_{CD} = N_{DE}
\end{equation*}

```python

nde, noe, nml = sy.symbols('nde noe nml')

SME = - RAy*(3*L) + nml*3*L*np.tan(a)
print('sum ME :', printsymSF(SME))
SMM = - RAy*(2*L) - nde*2*L*s(a)
print('sum MM :', printsymSF(SMM))
SFXe = nde*c(a) + noe*c(2*a) + nml
print('sum SFXe :', printsymSF(SFXe))

sol = solve({SME,SMM,SFXe},{nde, noe, nml})

NDE, NOE, NML = sol[nde], sol[noe], sol[nml]
print(f'NDE    = {NDE:.1f} kN')
print(f'NOE    = {NOE:.1f} kN')
print(f'NML    = {NML:.1f} kN')
```
