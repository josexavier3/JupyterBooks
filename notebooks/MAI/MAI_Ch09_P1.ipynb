## PRINCÍPIO DOS TRABALHOS VIRTUAIS

### Problema 1

Duas barras uniformes AB e BC, cada uma com massa igual a 4 kg, estão articuladas pelos pinos sem atrito B e C. A extremidade está assente numa superfície rugosa. Quando a força P atingir o valor de 90 N as barras encontram-se na situação de movimento iminente. Determine, para esta situação, qual o valor do coeficiente de atrito estático em A.

width="400">

#### Resumo

**Princípio dos Trabalhos Virtuais (PTV): párticula**

Considere uma particula sobre a qual é aplicada uma força externa. É possível calcular o trabalho realizado por esta força quando o seu ponto de aplicação se move ao longo de um caminho:

\begin{equation*}
\mathrm{d} U = \vec{F}\cdot\mathrm{d}\vec{r}
= |\vec{F}| |\mathrm{d}\vec{r}|\cos\theta
= (F\cos\theta)\mathrm{d}r
= F(\mathrm{d}r \cos\theta)
\end{equation*}

**Princípio dos Trabalhos Virtuais (PTV): estática**

O conceito de trabalho também pode ser aplicado à estática, onde não ocorrem deslocamentos. Neste contexto, os deslocamentos reais devem ser substituído por deslocamentos virtuais. Deslocamentos virtuais são deslocamentos (ou rotações) que são:

- fictícios, ou seja, não ocorrem na realidade,
- infinitesimalmente pequenos,
- geometricamente (cinematicamente) admissíveis, ou seja, são consistentem com as restrições (ligações com o exterior) do sistema.

Para distinguir os deslocamentos virtuais dos deslocamentos reais $\mathrm{d}\vec{r}$, denota-se deslocamentos virtuais como $\delta\vec{r}$.

Assim, o trabalho virtual realizado por forças ou momentos durante um deslocamento virtual é escrito como

\begin{equation*}
\delta U = \vec{F}\cdot\delta{d}\vec{r}
\quad\wedge\quad
\delta U = \vec{M}\cdot\delta\vec{\theta}
\end{equation*}

Este resultado pode ser generalizado. Para um sistema mecânico arbitrário, o trabalho virtual das forças externas e momentos externos é dado por:

\begin{equation*}
\delta U =
\sum\vec{F}_i\cdot\delta{d}\vec{r}_i
+
\vec{M}_i\cdot\delta\vec{\theta}_i
\end{equation*}

Para um sistema mecânico estar em equilíbrio, o PTV postula que é condição necessária e suficiente que a soma algébrica do trabalho virtual realizado por todas as forças e momentos externos que atuam sobre o corpo seja zero, para qualquer deslocamento (translação e rotação) virtual do sistema, cinematicamente admissível:

\begin{equation*}
\delta U = 0
\quad\therefore\quad
\textrm{o sistema mecânico está em equilíbrio}
\end{equation*}

_Deslocamento virtual_: Considere-se uma estrutura em equilíbrio. Nestas condições a estrutura está em repouso. Um deslocamento virtual é simplesmente um deslocamento admíssivel, embora hipotético, da estrutura.

_Trabalho virtual_: O trabalho virtual representa o trabalho que o carregamento externo (forças e momentos) realiza no sistema (em equilíbrio) para um dado deslocamento virtual admíssivel.

Aplicações:

- Cálculo de deslocamentos em estruturas isostáticas;
- Cálculo de forças em estruturas hiperestáticas;
- Equilíbrio de mecanismos.

### Resolução

**Análise do equilíbrio estático**

Considere-se o diagrama de corpo livre da estrutura (geometria e condições de fronteira, pela ação de forças diretamente aplicadas e ligação ao exterior)

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P1/MAI_Ch09_P1a.jpg"
width="400">

De notar que existem 4 incognítas no problema: duas reações no apoio duplo em C ($R_{Cx}, R_{Cy}$) e as forças geradas na condição de atrito seco em A ($N_{A}, F_{aA}$). No plano, como temos apenas 3 equações de equilíbrio estático ($\sum F_x = 0; \sum F_y = 0; \sum M_0 = 0$), a estrutura é dita hiperestática.

Contudo, pelo desenvolvimento algébrico das equações de equilíbrio é possível determinar parte das reações, momeadamente a força normal $N_A$ (pela equação do equilíbrio do momento fletor) e a componente da reação $R_{Cy}$, ficando a componente $R_{Cx}$ função do valor do coeficiente de atrito:

\begin{equation*}
\begin{cases}
\rightarrow(+)&:& \sum F_x = 0 \\
\uparrow(+) &:&\sum F_y = 0 \\
\circlearrowleft(+)&:&\sum M_C = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
F_a + C_x - P = 0 ~~\wedge~~ \large F_a = \mu_e N_A\\
N_A - (P1+P2) + C_y = 0 \\
P\cdot(200) + P_2\cdot(75) + P_1\cdot(225) - N_A\cdot(300) = 0
\end{cases}
\end{equation*}

resultando,

\begin{equation*}
\Leftrightarrow~
\begin{cases}
\mu_e N_A + C_x - P = 0 \\
N_A - (P1+P2) + C_y = 0 \\
-
\end{cases}
~\Leftrightarrow~
\begin{cases}
C_x = 90.0 - 99.24\mu~\textrm{N} \\
C_y = -20.8~\textrm{N}\\
N_A = 99.2~\textrm{N}
\end{cases}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import solve

mass = 4. # unit: kg
grav = 9.81 # unit: m/s2
P1 = mass*grav # unit N
print(f'P.AB = {P1:.2f} N')
P2 = mass*grav # unit N
print(f'P.BC = {P1:.2f} N')
P = 90. # unit: N
LX = 150. # unit: mm
LY = 200. # unit: mm

cx, cy, na, mu = sy.symbols('cx cy na mu')

# sumFx = ax - f
sumFx = mu*na + cx - P
print('\n   Sum Fx :',sumFx)
sumFy = na - P1 - P2 + cy
print('   Sum Fy :',sumFy)
sumMA = P*LY + P2*LX/2 + P1*(LX+LX/2) - na*(2*LX)
print('Sum sumMA :',sumMA)
sol = solve({sumFx,sumFy, sumMA},{cx, cy, na})

Cx = sol[cx]
print('\nCx = ',Cx,' N')
Cy = sol[cy]
print(f'Cy = {Cy:.1f} N')
NA = sol[na]
print(f'NA = {NA:.1f} N')
```

**Princípio dos Trabalhos Virtuais**

Considere-se a estrutura submetida a um conjunto de forças externas (diretamente aplicadas e forças de reação da ligação/fixação da estrutura). Considera-se que a estrutura está em repouso, pelo que os seus pontos estão fixos no espaço (não há aceleração ou forças de inércia na estrutura).

Considere-se agora, um deslocamento virtual da estrutura, hipotético mas admissível. O deslocamento da estrutura pode ser defino à custa de apenas um um grau de liberdade: por exemplo, o deslocamento infinitésimal da secção A, $\delta x_A
$, ou, de forma equivalente, a variação angular infinitéssimal em B, $\delta \theta_B$. Estes deslocamentos podem ainda ser relacionados com os deslocamentos virtuais no ponto D (centro de gravidade da barra BC aonde atuam o peso próprio da barra e uma força diretamente aplicada $P$).

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P1/MAI_Ch09_P1b.jpg"
width="400">

De notar que a força de reação $N_A$ não exerce trabalho pois faz um ângulo reto (90$^\circ$) com o deslocamento virtual horizontal $\delta x_A$. De notar ainda que a secção C está fixa, pelo que as forças de reação em C ($C_x, X_y$) não exercem trabalho virtual.

O trabalho virtual das forças aplicadas na estrutura em equilíbrio estático para o deslocamento virtual imposto (ver figura) é dado por:

\begin{equation*}
\begin{aligned}
\delta U = \sum_i \vec{F}_i\cdot\delta \vec{u} = 0
&~\Leftrightarrow~
\delta U_{F_a} + \delta U_{P_{1}} + \delta U_{P_{2}} + \delta U_P = 0\\
&~\Leftrightarrow~
F_a \delta x_A \cos(0^\circ)
+ P_{1}\,\delta y_E\cos(180^\circ)
+ P_{2}\,\delta y_D\cos(180^\circ)
+ P\delta x_D\cos(180^\circ) = 0\\
&\hspace{-.65cm}\textrm{como: $\cos(0^\circ)= 1$ ; $\cos(180^\circ) = -1$;
$P_1=P_2$; $\delta
y_E=\delta y_D$:}\\
&~\Leftrightarrow~
F_a \delta x_A - 2P_{1}\,\delta y_D
- P\delta x_D = 0
\end{aligned}
\end{equation*}

Por outras palavras, de notar que $F_a$ exerce um trabalho positivo (forças e deslocamentos têm o mesmo sentido) e $P_{1}$, $P_{2}$ e $P$ exerce um trabalho negativo (forças e deslocamentos têm sentidos opostos).

Por outro lado, as coordenadas da estrutura podem ser relacionadas pelas seguintes relações:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P1/MAI_Ch09_P1c.jpg"
width="400">


\begin{equation*}
\begin{cases}
x_D = -\displaystyle\frac{L}{2}\cos\theta \\
x_A = -2L\cos\theta \\
y_D = \displaystyle\frac{L}{2}\sin\theta
\end{cases}
\end{equation*}

onde L é o comprimento das barras AB e BC (com comprimentos iguais).
Derivando estas expressões relativamente a $\theta$ vem:

\begin{equation*}
\begin{cases}
\delta x_D = \displaystyle\frac{L}{2}\sin\theta\delta\theta
\\
\delta x_A = 2L\sin\theta\delta\theta \\
\delta y_D = \displaystyle\frac{L}{2}\cos\theta\delta\theta
\end{cases}
\end{equation*}

Aplicando o PTV resulta:

\begin{equation*}
\begin{aligned}
\delta U = 0 \quad\Leftrightarrow\quad &
F_a (2L\sin\theta\,\delta\theta)
- 2P_{1}\left(\frac{L}{2}\cos\theta\delta\theta\right)
- P\left(\frac{L}{2}\sin\theta\delta\theta\right) = 0
\\
\quad\Leftrightarrow\quad&
\left(F_a 2L\sin\theta
- P_{1}L\cos\theta
- \frac{PL}{2}\sin\theta\right)\delta\theta = 0
\quad\wedge\quad \delta\theta \ne 0 \\
\quad\Leftrightarrow\quad&
F_a = \frac{1}{4}(P + 2 P_2\cot\theta)\Bigg|_{\small\displaystyle\cot =
\frac{\cos\theta}{\sin\theta}} =
\frac{1}{4}\left(P + 2P_2\frac{1}{\tan\theta}\right)
\quad\wedge\quad P_1(=P_2) = mg \quad\wedge\quad
\theta = \arctan\left(\frac{400}{150}\right) = 69,4^\circ
\end{aligned}
\end{equation*}

```python
import numpy as np
fa,l,t = sy.symbols('fa l t')

xa = -2*l*sy.cos(t)
print('xA         = ',xa)
dxa = xa.diff(t)
print('dxA/dtheta = ',dxa)
dUFa = fa*dxa
print('dUFa       = ',dUFa)

yd = l/2*sy.sin(t)
print('\nyD         = ',yd)
dyd = yd.diff(t)
print('dyd/dtheta = ',dyd)
dUP2 = P2*dyd
print('dUP2       = ',dUP2)

xd = -l/2*sy.cos(t)
print('\nxD         = ',xd)
dxd = xd.diff(t)
print('dxd/dtheta = ',dxd)
dUP = P*dxd
print('dUP        = ',dUP)

dU = dUFa - 2*dUP2 - dUP
print('\ndU         = ',dU)

sol = solve(dU,fa)

Fa = sol[0]
print('\nFa         = ',Fa)

theta = np.arctan(2*LY/LX)
print('theta      = ','{:.2f}'.format(np.rad2deg(theta)),' graus')
NFa = Fa.subs(t,theta)
print('N[Fa]      = ','{:.2f}'.format(NFa),' N')
```

Pelo que,

\begin{equation*}
F_a = \mu_e N_A
~\Leftrightarrow~
\mu_e = \frac{F_a}{N_A} = 0.3
\end{equation*}

```python
nmu = NFa/NA
print('nmu      = ','{:.3f}'.format(nmu))

print('\nResumo forças reação:')
NCx = Cx.subs(mu,nmu)
print('Cx = ','{:.1f}'.format(NCx),'  N')
print('Cy = ','{:.1f}'.format(Cy),' N')
print('NA = ','{:.1f}'.format(NA),'  N')
```

##### Resolução alternativa para o cálculo de $N_A~\therefore~\mu$

Considere-se uma nova configuração virtual como representada na Fig. (iii.1):


<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P1/MAI_Ch09_P1d.jpg"
width="1000">

Desta configuração é possível obter as seguintes relações geométricas entre deslocamentos, desenvolvidos em relação à configuração de referência do problema:

\begin{equation*}
\begin{cases}
x_D = -\displaystyle\frac{L}{2}\cos\theta \\
y_D = \displaystyle\frac{L}{2}\sin\theta \\
y_E = \displaystyle\frac{L}{2}\sin\theta
\end{cases}
\end{equation*}

Derivando estas expressões relativamente a $\theta$ vem:

\begin{equation*}
\begin{cases}
\delta y_A = x_A\delta\theta = -2L\cos\theta\delta\theta\\
\delta y_E = \displaystyle\frac{L}{2}\cos\theta\delta\theta +
\frac{|\delta y_A|}{2} = \frac{3L}{2}\cos\theta\delta\theta\\
\delta y_D = \displaystyle\frac{L}{2}\cos\theta\delta\theta
\end{cases}
\end{equation*}

Aplica-se agora o PTV:

\begin{equation*}
\begin{aligned}
\delta U = \sum_i \vec{F}_i\cdot\delta \vec{u} = 0
&~\Leftrightarrow~
\delta U_{N_A} + \delta P_{1} + \delta P_{2} + \delta P = 0\\
&~\Leftrightarrow~
N_A \delta y_A
- P_{1}\,\delta y_D
- P_{2}\,\delta y_E
+ P\delta x_D = 0\\
&~\Leftrightarrow~
{N}_A \left(2L\cos\theta\delta\theta\right)
- P_{1}\left(\frac{L}{2}\cos\theta\delta\theta\right)
- P_{2}\left(\frac{3L}{2}\cos\theta\delta\theta\right)
+ P \left(\frac{L}{2}\sin\theta\delta\theta\right)= 0\\
&~\Leftrightarrow~
{N}_A 2L\cos\theta
- P_{1}\frac{L}{2}\cos\theta
- P_{2}\frac{3L}{2}\cos\theta
+ P \frac{L}{2}\sin\theta= 0
\quad\wedge\quad \delta\theta \ne 0\\
&~\Leftrightarrow~
N_A = \frac{P_1\cos\theta + 3P_2\cos\theta + P\sin\theta}{4\cos\theta}
\quad\wedge\quad
\mu_e = \frac{F_a}{N_A}
\end{aligned}
\end{equation*}

```python
NA_v2 = (P1 + 3*P2 + P*np.tan(theta))/4
print(f'NA (cal. 2)= {NA_v2:.1f} N')

nmu_v2 = NFa/NA_v2
print(f'nmu (cal. 2) = {nmu_v2:.3f}')
```
