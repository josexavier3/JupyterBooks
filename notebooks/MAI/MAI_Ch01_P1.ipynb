## Força e momento de uma força
### Problema 1


Três forças atuam em A como se indica na figura, de forma a equilibrar um vaso. Determine o vetor resultante das forças representadas, sabendo que a magnitude das mesmas é: $F_{AB} = 1308~N$, $F_{AC} = 763$ N, $F_{AD} = 708,5$ N  para $x = 2$ m e $z = 1,5$ m.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch01/P1/MAI_Ch01_P1.png"
width="800"/>

- Equações de equilíbrio para o ponto A,

\begin{equation*}
\vec{R} = \sum_i \vec{F}_i = \vec{F}_1 + \vec{F}_2 + \vec{F}_3 + \vec{F}_4
\end{equation*}

- Resultante das forças

Módulo,

\begin{equation*}
\vec{R} = \sqrt{\vec{R}\cdot\vec{R}} = \sqrt{\vec{R}_x^2 + \vec{R}_y^2} 
\end{equation*}

Direção,

\begin{equation*}
\tan(\theta) = \frac{\vec{R}_y}{\vec{R}_x} 
\end{equation*}


```python
import numpy as np
from numpy import cos, sin, sqrt, arctan
F1, F2, F3, F4 = 150, 80, 110, 100
angles_deg = [20, 30, 15]
ang1, ang2, ang4 = [np.deg2rad(angle) for angle in angles_deg]

Rx = F1*cos(ang2) - F2*sin(ang1) + F4*cos(ang4) 
print(f'Rx = {F1} ({cos(ang2):.2f}) - {F2} ({sin(ang1):.2f}) + {F4} ({cos(ang4):.2f}) = {Rx:.2f} N')

Ry = F1*sin(ang2) + F2*cos(ang1) - F3 - F4*sin(ang4) 
print(f'Ry = {F1} ({sin(ang2):.2f}) + {F2} ({cos(ang1):.2f}) - {F3} + {F4} ({sin(ang4):.2f}) = {Ry:.2f} N')

R = sqrt(Rx**2 + Ry**2)
print(f'R = {R:.2f} N')

t = np.rad2deg(arctan(Ry/Rx))
print(f't = {t:.2f} deg')
```

# Anexo

- Escolha de um referencial cartesiano associado ao ponto A;
- Exprimir os vetores em função das suas componentes do referêncial escolhido;

$\vec{AC}$

- pontos do segmento de recta:

$A(0,0,0)\quad\wedge\quad C(2,-6,3)$

\begin{equation*}
\vec{AC} = \vec{OC} - \vec{OA}  = (2,-6,3) = 2\vec{i} -6\vec{j} + 3\vec{k}
\end{equation*}

- Norma euclidiana de um vetor,

\begin{equation*}
||\vec{AC}|| = \sqrt{\vec{AC} \cdot \vec{AC}}
= \sqrt{x_{AC}^2 + y_{AC}^2 + z_{AC}^2}
\end{equation*}

- Versor do vetor $\vec{AC}$,

\begin{equation*}
\vec{\lambda}_{AC} = \frac{\vec{AC}}{||\vec{AC}||}
\end{equation*}

- Definição do produco interno ou produto escalar de dois vetores,

\begin{equation*}
\vec{a}\cdot\vec{b} = |a| |b| \cos\angle(\vec{a},\vec{b})   
\quad\therefore\quad \vec{a}\cdot\vec{a} = a²
\end{equation*}

- Cálculo matricial do produto interno ou escalar de dois vetores,

\begin{equation*}
\vec{a}\cdot\vec{b} = \sum_{i=1}^n a_i b_i = \{a\}^T \{b\}
\end{equation*}
