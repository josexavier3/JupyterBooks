### Resolução

#### a) A posição do centro geométrico da secção no referencial $S(O,x,y)$ indicado. ([_enunciatum_](#P5))

Considere-se a seguinte decomposição da superfície em estudo:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P5/MAI_Ch06_P5b.jpg"
width="800">

Os centroides das figuras do quadrado [1] e retângulo [2] são rapidamente identificáveis, pela sua simetria. No entanto, para a determinação da ordenada $y$ do centroide do semi-círculo mostrado na figura [3], recorre-se à tabela:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P5/MAI_Ch06_P5c.jpg"
width="700">

Pela decomposição apresentada, as coordenadas no referencial $S(0,x,y)$ do centroide da superfície original são dadas por:

\begin{equation*}
\begin{cases}
\overline{x} = b/2 = 60~\textrm{mm}
\qquad\Longleftarrow\quad\textrm{(por simetria)} \\
\overline{y} = \displaystyle\frac{\sum_{i} y_iA_i}{\sum_{i} A_i}
= \frac{y_1A_1 - y_2A_2 + y_3A_3}{A_1 - A_2 + A_3}
\end{cases}
\end{equation*}

Desenvolvendo numericamente para a coordenada $y$:

```python
import numpy as np

b = 120 # unit: mm
h = b
e = 10
r = 60

print('-------')
print('Area 1: ')
print('-------')
x1 = b/2; print('x1 = ','{:.1f}'.format(x1),' mm')
y1 = h/2; print('y1 = ','{:.1f}'.format(y1),' mm')
A1 = b*h; print('A1 = ','{:.1f}'.format(A1),' mm2')

print('\n-------')
print('Area 2: ')
print('-------')
x2 = b/2; print('x2 = ','{:.1f}'.format(x2),' mm')
y2 = e + (h-e)/2; print('y2 = ','{:.1f}'.format(y2),' mm')
A2 = (b-2*e)*(h-e);  print('A2 = ','{:.1f}'.format(A2),' mm2')

print('\n-------')
print('Area 3: ')
print('-------')
x3 = b/2; print('x3 = ','{:.1f}'.format(x3),' mm')
y3 = h + 4*r/3/np.pi; print('y3 = ','{:.1f}'.format(y3),' mm')
A3 = np.pi*r**2/2; print('A3 = ','{:.1f}'.format(A3),' mm2')

print('\n----------------------------')
print('Area Composta - (1)-(2)+(3): ')
print('----------------------------')
As = A1 - A2 + A3
print('A = ','{:.1f}'.format(As),' mm2')
Qys = x1*A1 - x2*A2 + x3*A3
print('Qy = ','{:.1f}'.format(Qys),' mm3')
Qxs = y1*A1 - y2*A2 + y3*A3
print('Qx = ','{:.1f}'.format(Qxs),' mm3')
xcoord = Qys/As; print('xcoord = ','{:.1f}'.format(xcoord),' mm')
ycoord = Qxs/As; print('ycoord = ','{:.1f}'.format(ycoord),' mm')
```

#### b) Os momentos de 2ª ordem da secção em relação aos eixos $x’$ e $y’$ paralelos a $x$, $y$ e com origem sobre o centro geométrico. ([_enunciatum_](#P5))

Para o cálculo do momento de inércia, segue-se um raciocínio similar ao da alínea anterior, _i.e._ divide-se a figura principal em três figuras mais simples e calculam-se os momentos de inércia para cada uma dessas três figuras mais simples (vide figura abaixo). Nesta abordagem é necessário atender aos diferentes sistemas de coordenadas locais tabelados, assim  o sistema de coordenadas com origem no centroide da figura composta.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P5/MAI_Ch06_P5d.jpg"

**Algoritmo de resolução**

- Em primeiro lugar calcula-se os momentos de inércia relativos aos eixos usados na tabela para uma determinada área.
- Se a origem dos eixos coordenadas na tabela não coincidir com o centroide dessa figura geométrica, será necessário um cálculo complementar para determinar o seu valor em relação ao um referencial local com origem no centroide, recorrendo ao teorema dos eixos paralelos. Esse é o caso para este exercício da
figura [3] do semi-círculo.
- Em seguida aplica-se o teorema dos eixos paralelos para a determinação do momento de inércia em relação a eixos que passam no centroide da figura original, partindo do conhecimento do momento de inércia em relação a eixos que passam no centroide local das áreas elementares. Esta transferência é feita recorrendo ao do teorema dos eixos paralelos. Neste processo, contudo, algumas simplificações ocorrem pelo fato do eixo $y$ ser um eixo de simetria.

**Quadrado**

Pretende-se determinar os momentos de inércia $I_{x'}$ e $I_{y'}$ da área do quadrado. Das tabelas contudo os momentos de inércia vêm dados em relação a um referencial local $I_{x''}$ e $I_{y''}$ (figra [1]).

```python
print('----------------')
print('Area 1: Quadrado')
print('----------------')
IxC1 = b*h**3/12
print(f'Ix (centroide local - tabela) = {IxC1:.3e} mm4')
IyC1 = h*b**3/12
print(f'Iy (centroide local - tabela) = {IyC1:.3e} mm4')
```

Dada a simetria da área em relação ao eixo $y$, têm-se: $I_{y'} = I_{y''}$.

```python
IySC1 = IyC1
print('Iy (centroide global - eixo simetria) = ','{:.3e}'.format(IySC1),' mm4')
```

Porém, para o momento de inércia em relação ao eixo $x'$ deve-se recorrer ao teorema dos eixos paralelos:

\begin{equation*}
I_{x'} = I_{x''} + A d_y^2
\end{equation*}

```python
dy1 = ycoord - y1
print('dy = ','{:.1f}'.format(dy1),' mm')
IxSC1 = IxC1 + A1*dy1**2
print(f'Ix (centroide global - teorema dos eixos paralelos) = {IxSC1:.3e} mm4')
```

**Retangulo**

Procede-se, agora, ao cálculo dos momentos de inércia relativamente aos eixos $x''y''$ do retângulo (figura [2]) e que correspondem aos momentos de inércia
relativamente aos eixos que passam pelo centroide da figura e estão tabelados.

```python
print('-----------------')
print('Area 2: Retangulo')
print('-----------------')
IxC2 = (b-2*e)*(h-e)**3/12
print('Ix (centroide local - tabela) = ','{:.3e}'.format(IxC2),' mm4')
IyC2 = (h-e)*(b-2*e)**3/12
print('Iy (centroide local - tabela) = ','{:.3e}'.format(IyC2),' mm4')
```

Notando novamente que que os eixos $y'$ e $y''$ são coincidentes, então:

```python
IySC2 = IyC2
print(f'Iy (centroide global - eixo simetria) = {IySC2:.3e} mm4')
```

Aplica-se o teorema dos eixos paralelos para determinar o momento de inércia relativamente ao eixo $x'$:

```python
dy2 = ycoord - y2
print('dy = ','{:.1f}'.format(dy2),' mm')
IxSC2 = IxC2 + A2*dy2**2
print(f'Ix (centroide global - teorema dos eixos paralelos) = {IxSC2:.3e} mm4')
```

**Semi-círculo**

Para calcular os momentos de inércia do semi-círculo relativamente aos eixos $x''y''$ (figura [3]), recorre-se à tabela:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P5/MAI_Ch06_P5e.jpg"
width="800">

Note-se que o eixo $x$ mostrado na tabela é coincidente com a base do semi-círculo. No entanto, para se determinar o momento de inércia relativamente ao eixo $x’$ (i.e. eixo horizontal que passa pelo centroide da figura) é necessário antes, calcular o momento de inércia relativamente ao centroide do semi-círculo. De notar que este cálculo não poderia ser obtido diretamente. Assim, aplica-se o teorema dos eixos paralelos e escreve-se:

\begin{equation*}
\begin{aligned}
I_\mathrm{x'''} & = I_\mathrm{x''} + Ad_1^2 \\
~\Leftrightarrow~&
I_\mathrm{x''} = I_\mathrm{x'''} - Ad_1^2
= \frac{\pi r^4}{8} -
\left(\frac{\pi r^2}{2}\right)
\left(\frac{4r}{3\pi}\right)^2
= \frac{(-32 + 9 \pi^2)r^4}{36 \pi}
\end{aligned}
\end{equation*}

e

\begin{equation*}
I_\mathrm{x'} = I_\mathrm{x''} + Ad_2^2
\end{equation*}

Sendo os eixos y’ e y’’ coincidentes:

\begin{equation*}
I_\mathrm{y'} = I_\mathrm{y''} = \frac{\pi r^4}{8}
\end{equation*}

```python
print('---------------')
print('Area 3: Círculo')
print('---------------')
dy3 = np.abs(ycoord - y3)
print('dy = ','{:.1f}'.format(dy3),' mm')
dyy = 4*r/3/np.pi
print('dyy = ','{:.1f}'.format(dyy),' mm')
IxC3 = np.pi/8*r**4 - A3*dyy**2
print('Ix (local centroid - table) = ','{:.3e}'.format(IxC3),' mm4')
IyC3 = np.pi/8*r**4
print('Iy (local centroid - table) = ','{:.3e}'.format(IyC3),' mm4')
IxSC3 = IxC3 + A3*dy3**2
print('Ix (global centroid) = ','{:.3e}'.format(IxSC3),' mm4')
IySC3 = IyC3
print('Iy (global centroid) = ','{:.3e}'.format(IySC3),' mm4')
```

**Area Composta**

Efectuando a soma dos momentos de inércia relativamente aos eixos $x'y'$ das três figuras vem:

```python
print('=============================')
print('Area Composta - (1)-(2)+(3): ')
print('=============================')
IxTOT = IxSC1 - IxSC2 + IxSC3
print('Ix (area total) = ','{:.3e}'.format(IxTOT),' mm4')
IyTOT = IySC1 - IySC2 + IySC3
print('Iy (area total) = ','{:.3e}'.format(IyTOT),' mm4')
```

#### c) O momento polar de inércia em relação à origem do referencial $S(O,x', y')$. ([_enunciatum_](#P5))

O momento polar de inércia da figura inicial em relação à origem do referencial $S(O,x', y')$ é dado por:

\begin{equation*}
J_0 = \int r^2 \mathrm{d}A\Bigg|_{r^2=x^2+y^2}
= \int x^2 \mathrm{d}A + \int y^2 \mathrm{d}A
= I_\mathrm{x',total} +
I_\mathrm{y',total}
\end{equation*}

#### e) Diga o que entende por eixos principais de inércia. Indique, justificando, se os eixos $x’y’$ são eixos principais de inércia para a secção dada. ([_enunciatum_](#P5))

O produto de inércia é uma medida da anti-simetria da distribuição de área de uma superfície em relação a um par de eixos. Os eixos principais de inércia definem um referencial sobre o qual o produto de inércia é nulo (existe simetria da distribuição de massa de um corpo em relação ao par de eixos). Os eixos de simetria são eixos principais de inércia.

O produto de inércia da superficie de área A relativamente aos eixos $x$ e $y$ é dado por:

\begin{equation*}
I_{xy} = \int xy \, \mathrm{d}A
\end{equation*}

Sempre que o produto de inércia de uma superfície em relação a um par de eixos xy se anula, esse par de eixos toma a designação de eixos principais de inércia da superfície.

Quando um ou ambos os eixos x e y são eixos de simetria da superfície de área A, o produto de inércia é nulo. Logo esses eixos são eixos principais de inércia, como é o caso dos eixo x’y’ neste problema.

Os eixos principais de inércia de uma superfície com origem sobre o centróide dessa superfície designam-se por eixos principais centrais de inércia da superfície.

#### Anexo

**Momentos e produtos de inércia**

Considere-se agora o segundo momento de área ou momento de inércia. Este representa uma medida relativa da forma como a área está distribuída em relação a um eixo de interesse. O momento de inércia da área A em relação aos eixos $x$ e $y$ é dado, respetivamente, por:

\begin{equation*}
I_x = \int_A y^2\,\mathrm{d}A~,
\quad
I_y = \int_A x^2\,\mathrm{d}A
\end{equation*}

Em que $x$ e $y$ são as coordenadas de um elemento de área $\mathrm{d}A$ da secção. Os momentos de inércia são sempre quantidades positivas, na medida em que as coordenadas $x$ e $y$ aparecem ao quadrado.

Por sua vez, o produto de inércia da área A em relação aos eixos $x$ e $y$ é definido por:

\begin{equation*}
I_{xy} = \int_A xy\,\mathrm{d}A
\end{equation*}

Neste caso, cada elemento de área $\mathrm{d}A$ é multiplicado pelas suas coordenadas $x$ e $y$. O produto de inércia de área em relação a um par de eixos é nulo, quando um dos eixos é eixo de simetria.

**Teorema dos eixos paralelos**

O momento de inércia de área em relação ao um determinado eixo pode ser relacionado com o momento de inércia em relação a um eixo paralelo que passa no centroide da secção em estudo, pelo teorema dos eixos paralelos.

Para o desenvolvimento deste teorema, considere-se o elemento de área $\mathrm{d}A$. Seja $\overline{x}$ e $\overline{y}$ as coordenadas do centroide de área, e o referencial $x-y$ paralelo a estes eixos.
As distâncias entre os pares de eixos são designadas por $\mathrm{d}x$ e $\mathrm{d}y$. O momento de inércia em relação a $x$ pode ser expresso por:

\begin{equation*}
I_x = \int_A (y+\overline{y})^2\,\mathrm{d}A =
 \int_A y^2\,\mathrm{d}A+2\overline{y}\int_A y\,\mathrm{d}A
+\overline{y}^2\int_A \mathrm{d}A
\end{equation*}

O primeiro integral no segundo membro representa o momento de inércia em relação ao eixo $\overline{x}$:

\begin{equation*}
I_{\overline{x}} =
\int_A y^2\,\mathrm{d}A
\end{equation*}

Por outro lado, como $y$ é medido em relação ao centroide $\overline{x}$, o segundo integral será zero:

Por outro lado, como $y$ é medido em relação ao centroide $\overline{x}$, o segundo integral será zero:

Desta forma,

\begin{equation*}
I_x = I_{\overline{x}} + A\overline{y}^2
= I_{\overline{x}} + Ad_y^2
\end{equation*}

De forma similar,

\begin{equation*}
I_y = I_{\overline{y}} + A\overline{x}^2
= I_{\overline{y}} + Ad_x^2
\end{equation*}
