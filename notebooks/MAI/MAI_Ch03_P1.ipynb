#### b) Calcule a intensidade da força $\vec{T}$ que é necessário aplicar para que a viga fique em equilíbrio estático quando $\alpha = 30^\circ$.

A aplicação das equações de equilíbrio requer a especificação de todas as forças conhecidas e desconhecidas que atuam sobre o corpo. A melhor maneira de quantificar todas essas forças é desenhar o diagrama de corpo livre do corpo.

Qualquer sistema de forças, $F_1, F_2, \ldots, F_n$ pode ser substituído por um sistema força-binário equivalente aplicado num dado ponto $O$, constituído pelos vetores $\vec{R}$ e $M_0^R$. Quando as forças que actuam sobre um corpo podem ser reduzidas, num ponto arbitrário $O$, a um sistema força-binário constituído por uma força nula e um binário nulo, diz-se que o corpo está em equilíbrio
estático.

As condições necessárias e suficientes para garantir o equilíbrio são:

\begin{equation*}
\begin{cases}
\sum\vec{F} = \vec{0}\\
\sum\vec{M}_0 = \sum\vec{r}\times\vec{F} = \vec{0}
\end{cases}
\end{equation*}

Decompondo a força e o momento nas suas componentes cartesianas, estas duas equações vectoriais conduzem às seguintes seis equações escalares:

\begin{equation*}
\begin{cases}
~\sum F_x = 0\\
~\sum F_y = 0\\
\sum M_0 = 0 \\
\end{cases}
\end{equation*}

Para escrever as equações de equilíbrio de um corpo rígido é essencial identificar todas as forças aplicadas no corpo.  Essas forças devem constar do diagrama de corpo livre.

Comece-se por considerar o diagrama de equilíbrio da viga ABC, nomeadamente com o intuito de determinar a força $\vec{R}_1$ aplicada na roldana em D.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P1/MAI_Ch03_P1b.jpg"
width="700">

Desenvolvendo o sistema de equações de equilíbrio (considere-se convenientemente
o ponto C como polo para o equilíbrio do momento) vem:

\begin{equation*}
\begin{cases}
\sum\vec{F} = \vec{0}\\
\sum\vec{M}_C = \vec{0}
\end{cases}\Leftrightarrow
\begin{cases}
\vec{R}_1 + \vec{P} + \vec{R}_\textrm{C} = \vec{0} \\
\vec{r}_\textrm{CA}\times\vec{R}_1 + \vec{r}_\textrm{CB}\times\vec{P} =
\vec{0}
\end{cases}
\end{equation*}

\begin{equation*}
\Leftrightarrow
\begin{cases}
\rightarrow(+)&~\sum F_x=0~:& R_{Cx} = 0
\\
\uparrow(+)&~\sum F_y=0~:& R_1 - P + R_{Cy} = 0
\\
\circlearrowleft(+)&\sum M_A=0:& (-l_\textrm{AC} \vec{\imath})\times(R_1\,
\vec{\jmath}) +
(-l_\textrm{BC} \vec{\imath})\times(-P\,\vec{\jmath}) =
\vec{0}
\end{cases}
\end{equation*}

\begin{equation*}
\Leftrightarrow
\begin{cases}
R_{Cx} = 0 \\
R_{Cy} = P - R_1 \\
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
-l_\textrm{AC} & 0 & 0 \\
0 & R_1 & 0
\end{vmatrix}+
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
-l_\textrm{BC} & 0 & 0 \\
0 & -P & 0
\end{vmatrix} = \vec{0}
\end{cases}
\end{equation*}

\begin{equation*}
\Leftrightarrow
\begin{cases}
- \\
- \\
(-l_\textrm{AC}) R_1 \vec{k} + (-l_\textrm{BC})(-P) \vec{k} = \vec{0}
\end{cases}
\Leftrightarrow
\begin{cases}
R_{Cx} = 0 \\
R_{Cy} = P - R_1 \\
R_1 = \displaystyle\frac{l_\textrm{BC} }{l_\textrm{AC}} P
\end{cases}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

# unit: N, m
P, lAC, lBC = 180,  0.9,  0.6 # unit: m

print('code 1 (sympy.solve) ---------------- ')
rcx, rcy, r1 = sy.symbols('rcx rcy r1')

SFX = rcx
SFY = rcy - P + r1
SMC = -lAC*r1 + lBC*P
sol = solve({SFX,SFY,SMC},{rcx, rcy, r1})

RCx = float(sol[rcx]); print('RCx = ','{:.1f}'.format(RCx),' N')
RCy = float(sol[rcy]); print('RCy = ','{:.1f}'.format(RCy),' N')
R1 = float(sol[r1]);   print(' R1 = ','{:.1f}'.format(R1),' N')

print('\ncode 2 (sympy.physics.mechanics) ---------------- ')

from sympy.physics.mechanics import ReferenceFrame, cross
S = ReferenceFrame('S') # ||x|| = ||y|| = ||z|| = 1
r11 = sy.symbols('r11')

vec_rCA = -lAC*S.x
vec_R1 = r11*S.y

vec_rCB = -lBC*S.x
vec_P = -P*S.y

expMc = cross(vec_rCA,vec_R1) + cross(vec_rCB,vec_P)
print('Exp Mc = ',expMc)
EQMc = expMc.args[0][0].args[2][2]
sol = solve({EQMc},{r11})
R11 = float(sol[r11])
print(' R1 = ','{:.1f}'.format(R11),' N')
```

Analisando agora o equilíbrio da roldana em D (a roldana não absorve momentos, pelo que apenas se analisa o equilíbrio pelo $\sum\vec{F}=\vec{0}$;  por outro lado, assume-se que as forças são concorrentes no ponto D,  desprezando o raio, massa e atrito da roldana):

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P1/MAI_Ch03_P1e.jpg"
width="700">

\begin{equation*}
\begin{cases}
\sum\vec{F} = \vec{0}
\end{cases}\Leftrightarrow
\begin{cases}
2\vec{T} + \vec{R}_1 + \vec{T} = \vec{0}
\end{cases}
\end{equation*}

\begin{equation*}
\Leftrightarrow
\begin{cases}
\rightarrow(+)&x:& -2T\cos\beta + T \cos\alpha = 0 \\
\uparrow(+)&y:& 2T\sin\beta + T \sin\alpha - R_1 = 0
\end{cases}
\end{equation*}

\begin{equation*}
\Leftrightarrow
\begin{cases}
\cos\beta = \displaystyle\frac{1}{2}\cos\alpha \\
T = \displaystyle\frac{R_1}{2\sin\beta + \sin\alpha}
\end{cases}, \quad \alpha=30^\circ
\end{equation*}

```python
t, b = sy.symbols('t b')
alf = np.deg2rad(30) # unit: radianos

EQX = -2*t*sy.cos(b) + t*np.cos(alf)
EQY = 2*t*sy.sin(b) + t*np.sin(alf) - R1
sol = solve({EQX,EQY},{t, b})

T = float(sol[0][t]); print('T = ','{:.1f}'.format(T),' N')
angBet = float(sol[0][b]); angBetDeg = np.rad2deg(angBet)
print('angBeta = ','{:.1f}'.format(angBetDeg),' graus')
```

#### Apêndice

Em várias aplicações, nem sempre é possível tratar um corpo como uma partícula. Nomeadamente, quando as forças atuantes num corpo não são concorrentes num ponto, as dimensões do corpo e os pontos de aplicação das forças têm de ser considerados. As forças que atuam num corpo rígido podem ser classificadas em forças externas e forças internas (estas forças interiores são responsáveis por manter unidas as diferentes partículas que constituem o corpo rígido). Neste capítulo dar-se-á sobretudo ênfase às forças externas. Estas representam a ação de outros corpos sobre o corpo rígido em análise.

**Diagrama de Corpo Livre**

- Desenho simplificado do corpo ou da parte do corpo que se pretende estudar, desligado dos apoios e separado de todos os outros corpos.

- Com todas as forças exteriores aplicadas sobre o corpo.  Essas forças representam a ação dos outros corpos, do peso próprio (aplicado no centro de gravidade) e dos apoios e estão aplicadas no ponto de contacto com o corpo.

- As intensidades, direções e sentidos das forças exteriores conhecidas deverão estar claramente representadas.

- As forças exteriores desconhecidas consistem geralmente em reações nos apoios.

- O diagrama de corpo livre deve incluir as dimensões necessárias à resolução do problema.
