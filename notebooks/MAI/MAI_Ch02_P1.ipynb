### Problema 1

Um caixote de 2670 N ($\approx$ 267 kg) é suportado por vários conjuntos de roldanas, representados na figura. Determine, para cada conjunto, a força de tração que é necessário exercer na corda.

#### Questões de análise

> Dos cinco sistemas de roldanas apresentados, qual é o que conduz à menor força a aplicar, $\vec{T}$, para equilibrar o peso do objeto?

> Existem arranjos de roldanas equivalentes?

#### História

O mecanismo da polia é uma invenção que representou um grande avanço tecnológico na Antiguidade. A sua invenção é atribuída a Arquimedes (287-212 a.C.), na Grécia Antiga. O mecanismo consiste numa associação de polias móveis em relação a pontos de fixação. O enunciado do problema exemplifica arranjos possível para esse tipo de aparato.

A história conta que Arquimedes teria demonstrado ao rei Hierão o seu engenho, movendo sozinho, sobre a areia da praia, um navio repleto de passageiros e cargas, um feito à partida inimaginável.

#### Hipóteses

Várias aplicações práticas de engenharia envolvem forças coplanares. Exemplos incluem forças que atuam em roldanas (polias), lançamento de projéteis e equilíbrio de objetos planos.

A roldana é um mecanismo que permite, por um lado, alterar a direção da força (pela alteração da direção do cabo) e, por outro lado, modificar a magnitude da força que se transmite através dela, nomeadamente por um arranjo adequado de várias polias.

As roldanas podem ser usadas por exemplo, para reduzir a força aplicada para mover um determinado objeto com um dado peso.

A metodologia de análise consiste em reduzir a roldana a uma partícula e representar sobre esta todas as forças externas que nela atuam. Este diagrama de forças é desinado por diagrama de corpo livre (DCL) da partícula.

De notar ainda que a força de tração nos dois troços do cabo que abraçam uma roldana, são considerados de igual intensidade. Naturalmente, esta aproximação é válida apenas, por um lado, se for assumida a ausência de atrito e, por outro lado, considerando a massa da roldana nula (para se poder desprezar o seu momento de inércia).


### Resolução

#### (a)

O diagrama de corpo livre do sistema caixote-roldana que suporta o peso de 2, 67 kN resulta num sistema de forças como ilustrado na figura

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P1/MAI_Ch02_P1a.jpg"
style="max-width: 100%;"/>

De salientar que a força exercida em ambos os tramos do cabo são consideradas iguais em módulo ($T$). Assumindo que o sistema de força aplicadas à partícula (roldana) está em equilíbrio, resulta (neste caso, a resultante das forças reduz-se a uma equação escalar do equilíbrio as forças segundo o eixo vertical, $y$):

\begin{equation*}
\sum \vec{F} = \vec{0}
\Leftrightarrow
\sum F_y = 0
\Leftrightarrow
2T - P = 0
\Leftrightarrow T=\frac{P}{2}~\mathrm{[N]}
\end{equation*}

De notar que este mecanismo, com uma única roldana, permite reduzir para metade do peso do caixote o valor da força a aplicar, para suster (equilibrar) o objeto.

```python
P = 2670 # unit: Newton

print('a)')
# code 1:
print('code 1:')
T_a = P/2
print('T = ','{:.1f}'.format(T_a),' N')

# code 2:
print('\ncode 2:')
import sympy as sy
from sympy.solvers import solve

t, p = sy.symbols('t p')
SFY = 2*t - p
print('sum Fy :', SFY)
sol = solve({SFY}, {t})

T2 = sol[t]
print('T = ', T2,' kN')
NT2 = float(T2.subs({(p,P)})) # float: converte n. inteiro em virgula flutuante
print('N[T] = ', NT2,' kN')
```

Na roldana superior (sem atrito e massa) a força em ambos os extremos do cabo é

agora exercida de cima para baixo.


#### (c)

Neste arranjo, existem duas roldanas, uma fixa (superior) e outra móvel (inferior). Na figura ilustra-se o diagrama de corpo livre individual da roldana móvel. O diagrama ilustra as três forças  de tração exercidas pelos cabos e a força exercída pelo caixote, pela ação da gravidade.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P1/MAI_Ch02_P1c.jpg"
style="max-width: 100%;"/>

Do equilíbrio de forças aplicadas conclui-se:

\begin{equation*}
\sum F_y = 0
\Leftrightarrow 3T - P = 0
\Leftrightarrow T=\frac{P}{3}~\textrm{[N]}
\end{equation*}

Neste sistema de roldanas a força aplicada para equilibrar o peso do caixote é três vezes inferior.

```python
print('\nc)')

# code 1:
print('\ncode 1:')
T_c = P/3
print('T(c) = ','{:.1f}'.format(T_c),' N')

# code 2:
print('\ncode 2:')

tc = sy.symbols('tc')
SFY = 3*tc - p
print('sum Fy :', SFY)
sol = solve({SFY}, {tc})

T2c = sol[tc]
print('T(c) = ', T2c,' kN')
NT2c = float(T2c.subs({(p,P)})) # float: converte n. inteiro em virgula
# flutuante
print('N[T] = ', NT2c,' kN')
```

#### (e)

O diagrama de corpo livre da roldana inferior permite concluir que existem duas forças de intensidade $T$ mais uma força de intensidade $2T$ da roldana intermédia, que equilibram a força associada ao peso próprio do objeto.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P1/MAI_Ch02_P1e.jpg"
style="max-width: 100%;"/>

Da expressão do equilíbrio estático resulta:

\begin{equation*}
\sum F_y = 0 \Leftrightarrow T+2T+T=P
\Leftrightarrow T=\frac{P}{4}~\mathrm{[N]}
\end{equation*}

A força a aplicar será quatro vezes inferior ao peso do caixote.

```python
print('\ne)')

# code 1:
print('\ncode 1:')
T_e = P/4
print('T = ','{:.1f}'.format(T_e),' N')

# code 2:
print('\ncode 2:')

te = sy.symbols('te')
SFY = 4*te - p
print('sum Fy :', SFY)
sol = solve({SFY}, {te})

T2e = sol[te]
print('T(e) = ', T2e,' kN')
NT2e = float(T2e.subs({(p,P)})) # float: converte n. inteiro em virgula
# flutuante
print('N[T] = ', NT2e,' kN')
```
