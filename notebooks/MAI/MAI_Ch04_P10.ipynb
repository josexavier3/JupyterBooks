A figura representa uma estrutura metálica onde se encontra suspenso um cilindro de 300 N por um cabo em D. A estrutura está apoiada por uma junta articulada em A e suspensa no ponto B através do cabo BC. Despreze o peso próprio da estrutura.

a)	Represente o diagrama de corpo livre da estrutura.

b)	Determine a intensidade da força suportada pelo cabo.

c)	Determine as componentes da reação no apoio A.

### Resolução

#### Forças do diagrama de corpo livre

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch04/P10/MAI_Ch04_P10_i.png"
width="500">

- Forças e momentos de reação em A:

\begin{equation*}
\vec{R}_A = A_x~\vec{i} + A_y~\vec{j} + A_z~\vec{k}
\quad\wedge\quad
\vec{M}_A = M_y~\vec{j} + M_z~\vec{k}
\end{equation*}

- Força no cabo BC:

\begin{equation*}
\vec{T}_{BC} = T~\cdot~\vec{\lambda}_{BC} = T~\cdot~\frac{\vec{BC}}{||\vec{BC}||}
\end{equation*}

- Força em D:

\begin{equation*}
\vec{F} = -F_D~\vec{k}= -300~\vec{k}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
from sympy.physics.mechanics import ReferenceFrame, cross

A = [0.,0.,0.]
B = [-4.,6.,0.] # unit: m
C = [-1.,0,2.]
D = [-2.,6.,0.]

def vetor(a,b):
    return  np.subtract(b,a)

def versor(a,b):
    v1 = np.subtract(b,a)
    modv1 = np.sqrt(np.dot(v1,v1))
    nv1 = v1/modv1
    return v1, modv1, nv1

vAD = vetor(A,D)
print(f'Position vectors ------------ : \n vAD = {vAD} [m]')

vAB = vetor(A,B)
print(f' vAB = {vAB} [m]')

vBC, modBC, nBC = versor(B,C)
print(f' vBC = {vBC} | norma de BC = {modBC:.3f} | versor dir. BC = {nBC} [m]')


S = ReferenceFrame('S') # ||x|| = ||y|| = ||z|| = 1
t, ax, ay, az, my, mz = sy.symbols('t ax ay az my mz')

F = 300 # unit: N

vecAD = vAD[0]*S.x + vAD[1]*S.y + vAD[2]*S.z
vecF = -F*S.z
vecAB = vAB[0]*S.x + vAB[1]*S.y + vAB[2]*S.z
vecBC = nBC[0]*S.x + nBC[1]*S.y + nBC[2]*S.z
vT = t*vecBC
vA = ax*S.x + ay*S.y + az*S.z
MA = my*S.y + mz*S.z

print(f'Force and moment vectors ------------ : \n vecF = {vecF} [N]')
print(f' vecT = {vT} [N]')
print(f'   vA = {vA} [N]')
print(f'   MA = {MA} [N.m]')
```

#### Equilíbrio estático

A condição de equilíbrio de um sólido rígido no espaço pressupõe:

\begin{equation*}
\bullet\quad\sum \vec{F} = \vec{0}~\Leftrightarrow~
\begin{cases}
\rightarrow(+)&x:&\sum F_x = 0 \\
\uparrow(+)&y:&\sum F_y = 0 \\
\swarrow(+)&z:&\sum F_z = 0 \\
\end{cases}
\end{equation*}

e

\begin{equation*}
\bullet\quad\sum \vec{M}_A = \vec{0}~\Leftrightarrow~
\begin{cases}
\vec{k}\times\vec{\jmath} = \vec{\imath}~(+)&:&\sum M_{Ax} = 0 \\
\vec{k}\times\vec{\imath} = \vec{\jmath}~(+)&:&\sum M_{Ay} = 0 \\
\vec{\imath}\times\vec{\jmath} = \vec{k}~(+)&:&\sum M_{Az} = 0 \\
\end{cases}
\end{equation*}


- Expressão do momento em torno do ponto A,

\begin{equation*}
\sum M_A = 0
\quad\therefore\quad
\vec{r}_{AD} \times \vec{F}_D + \vec{r}_{AB} \times \vec{T} + \vec{M}_A = \vec{0}
\end{equation*}

\begin{equation*}
\vec{r}_{AD} \times \vec{F}_D + \vec{r}_{AB} \times (T \cdot \vec\lambda_{BC}) + \vec{M}_A = \vec{0}
\end{equation*}

```python

# Sum forces (vector equation):
print('FORCES :::::::::::::::::::::')
ExpF = vecF + vT + vA
print('Sum F  : ',ExpF)

Eq1 = ExpF.args[0][0].args[2][0] # t*nBC[0] + ax
Eq2 = ExpF.args[0][0].args[2][1] # t*nBC[1] + ay
Eq3 = ExpF.args[0][0].args[2][2] # t*nBC[2] -F + az

print(f'\nsum Fx : {Eq1} \nsum Fy : {Eq2} \nsum Fz : {Eq3}')
```

```python
# Sum moments (vector equation):
print('MOMENTS :::::::::::::::::::::')
CP1 = cross(vecAD,vecF); print('Cross Prod. F: ',CP1)
CP2 = cross(vecAB,vT); print('Cross Prod. T: ',CP2)
expMA = CP1 + CP2 + MA
print('Sum MA = ',expMA)

Eq4 = expMA.args[0][0].args[2][0]
Eq5 = expMA.args[0][0].args[2][1]
Eq6 = expMA.args[0][0].args[2][2]
print(f'\nsum Mx : {Eq4} \nsum My : {Eq5}\nsum Mz : {Eq6}')
```

```python

# solving a set of 6 scalar equations

sol = solve({Eq1,Eq2,Eq3,Eq4,Eq5,Eq6},{t, ax, ay, az, my, mz})

T   = float(sol[t]);  print(f'  T = {T:.1f} N')
RAx = float(sol[ax]); print(f'RAx = {RAx:.1f} N')
RAy = float(sol[ay]); print(f'RAy = {RAy:.1f} N')
RAz = float(sol[az]); print(f'RAz = {RAz:.1f} N')
MAy = float(sol[my]); print(f'MAy = {MAy:.1f} N.m')
MAz = float(sol[mz]); print(f'MAz = {MAz:.1f} N.m')
```
