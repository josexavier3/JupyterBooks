#### Problema 17

Na figura está representada uma placa fina semicircular ABC, que se encontra apoiada em A por uma rótula esférica, e nas extremidades B e C pelos cabos BD e CD respetivamente. A superfície de topo dessa placa está no plano horizontal $xy$. Nessa placa é aplicada uma força vertical (segundo z) de 300 N, e a sua linha de ação passa pelo centro de gravidade da placa ($d$ = 0,637 m). Despreze o peso próprio da placa.

a) Represente o diagrama de corpo livre da placa e determine os valores da força em cada cabo e as componentes da reação em A.

b) Considere agora que é adicionada uma força na direção $x$ aplicada num ponto do bordo BC da placa com intensidade 300 N e a uma distância $h$ do eixo $x$. Determine de forma justificada o valor da distância h de forma a manter o equilíbrio estático da placa.
