#### Problema 11

A Figura estrutura está apoiada em A e E por apoios ﬁxos. Despreze o peso da estrutura.

a) Represente o diagrama de corpo livre da viga AB e da viga EDC.

b) Veriﬁque que as componentes da reação em A são: Ax = 12 kN e Ay = 19 kN.

c) Determine as funções dos esforços internos para a viga AB e trace os respetivos diagramas.

Inclua no diagrama a localização da secção da viga com o momento ﬂetor máximo e
determine o respetivo valor.

## Resolução

```python
import numpy as np
import sympy as sp

h: float = 1.5 # unit: m
v: float = 2.0
F: float = 5.0 # unit: kN
dist: float = 4.0 # unit: kN/m

angEAD = np.arctan(h/v)
print(f'angEAD = {angEAD*180/np.pi:.2f}º')

angDAB = np.arctan(v/h)
print(f'angDAB = {angDAB*180/np.pi:.2f}º')

ax, ay, ex, ey, fad, fbc = sp.symbols('ax ay ex ey fad fbc')

# barra AB
SFX1 = sp.Eq(ax + fad*np.cos(angDAB), 0)
SFY1 = sp.Eq(ay - fad*np.sin(angDAB) - (dist*h)/2 - (dist*h)/2 - fbc, 0)
SMA1 = sp.Eq(-(dist*h/2)*(2*h/3) - (dist*h/2)*(h + h/3) - fbc*(2*h), 0)

# barra EDC
SFX2 = sp.Eq(ex - fad*np.cos(angDAB), 0)
SFY2 = sp.Eq(ey + fad*np.sin(angDAB) - F + fbc, 0)
SME2 = sp.Eq(fad*np.sin(angDAB)*(h) - F*(2*h) + fbc*(2*h), 0)

sol = sp.solve([SFX1, SFY1, SMA1, SFX2, SFY2, SME2], (ax, ay, ex, ey, fad, fbc))

formatted_sol = {key: f'{value:.2f}' for key, value in sol.items()}
print(f'sol = {formatted_sol}')
```
