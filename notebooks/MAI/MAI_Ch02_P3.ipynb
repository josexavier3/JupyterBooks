# Resolução

Pelo equilíbrio da roldana em C.

Nota: a tração $T_{AC} = P$

```python
from numpy import cos, sin, deg2rad
from sympy import symbols, solve

P = 750. # N
ang55 = deg2rad(55.)
ang25 = deg2rad(25.)

tabc, q = symbols('tabc, q')

SFx_roldC = -tabc*cos(ang55) + tabc*cos(ang25) - P*cos(ang55)
SFy_roldC = tabc*sin(ang55) + tabc*sin(ang25) + P*sin(ang55) - q
sol = solve({SFx_roldC, SFy_roldC}, {tabc, q})
TABC = sol[tabc]
Q = sol[q]
print(f'TABC = {TABC:.1f} N')
print(f'Q = {Q:.1f} N')
```
