### Problema 1

Um tubo de alumínio com 750 mm de comprimento e área da secção transversal igual a 1200 mm $²$ é soldado a um suporte fixo A e a uma tampa rígida B. A haste de aço EF, com 19 mm de diâmetro, é soldada à tampa B. Sabendo que os módulos de elasticidade do aço e do alumínio são respetivamente iguais a 200 GPa e a 70 GPa, determine:

a) A energia de deformação total do sistema quando $P = 36$ kN.

b) A correspondente densidade da energia de deformação no tubo CD e na haste EF.

### Resumo

#### Teorema do trabalho-energia para um carregamento único

O princípio da conservação de energia estabelece que a energia de um sistema fechado nunca é criado ou destruído, mas sempre transformado de um estado para outro. Por exemplo, o trabalho de uma força externa atuando sobre um corpo deformável é transformado em energia de deformação elástica interna (assumindo que nenhuma energia é perdida na forma de calor).
Assim, a energia de deformação $U$ é igual em magnitude ao trabalho externo $W$:

\begin{equation*}
U = W
\end{equation*}

A energia de deformação elástica em vigas prismáticos carregadas axialmente pode ser determinada a partir de,

\begin{equation*}
U = \frac{P² L }{2AE}
\end{equation*}

Concretamente, para elementos axiais (simples ou construídos por vários membros axiais prismáticos), a energia de deformação total no membro ou estrutura pode ser calculada por,

\begin{equation*}
U = \sum_{i=1}^n \frac{P_i² L_i}{2A_iE_i}
\end{equation*}

##### a) A energia de deformação total do sistema quando $P = 36$ kN.

A energia de deformação total do sistema mecânica constituido pelas peças de alumínio e aços vem,

\begin{equation*}
U = \sum_{i=1}^2 \frac{P_i^2L_i}{2A_iE_i}
= \left.\frac{P^2L}{2AE}\right|_{\textrm{Al}}
+ \left.\frac{P^2L}{2AE}\right|_{\textrm{Acco}}
\end{equation*}

```python
def U(p,l,a,e):  return p**2*l/2/a/e

U_Al = U(d.P,Al.l,Al.A,Al.E)
print(f'U_Al = {U_Al:.3f} [J = N.m]')

U_aco = U(d.P,aco.l,aco.A,aco.E)
print(f'U_aco = {U_aco:.3f} [J = N.m]')

U_tot = U_Al + U_aco
print(f'U_tot = {U_tot:.3f} [J = N.m]')
```

#####  b) A correspondente densidade da energia de deformação no tubo CD e na haste EF.

Para normalizar a solução eliminando o efeito das dimensões, calcula-se a energia de deformação por unidade de volume. A densidade da energia de deformação pode ser obtida por,

\begin{equation*}
u = \frac{U}{V} \quad\mathrm{[J/m³]}
\end{equation*}

```python
Al.vol = Al.A*Al.l
u_tubo = U_Al/Al.vol
print(f'u_tubo = {u_tubo:.1f} [J/m³]')

aco.vol = aco.A*aco.l
u_aco = U_aco/aco.vol
print(f'u_aco = {u_aco:.1f} [J/m³]')
```
