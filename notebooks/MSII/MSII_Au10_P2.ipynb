### Problema 2

O anel D pesando 330 N é libertado da posição de repouso indicada e atinge a placa ligada à extremidade C da barra vertical ABC. Sabendo que $E = 200$ GPa para ambos os troços da barra, determine a distância h para a qual a tensão máxima na barra é 200 MPa.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au10/P2/MSII_Au10_P2.png"
style="width: 25%;"/>

### Resolução

\begin{equation*}
U = \frac{P²L_{AB}}{2A_{AB}E} + \frac{P²L_{BC}}{2A_{BC}E} 
\end{equation*}


```python
import numpy as np
from typing import NamedTuple

# Dados
# unidades SI: m, s
class dados(NamedTuple):
    P : float # m
    E : float
    SigMax : float
    lAB : float
    dAB : float
    lBC : float
    dBC : float
    
d = dados(330., 200.e9, 200.*1e6, 2.4, 45.e-3, 1.2, 38.e-3)

def AreaCirc(d): return np.pi*d**2/4

AAB = AreaCirc(d.dAB)
ABC = AreaCirc(d.dBC)
print(f'AAB = {AAB:.2e} m² | ABC = {ABC:.2e} m²')

P_ = d.SigMax*ABC
print(f"P (Sigma Max) = {P_:.1f} N")

def U(p,l,a,e): return p**2*l/(2*a*e)

UAB = U(P_, d.lAB, AAB, d.E)
UBC = U(P_, d.lBC, ABC, d.E)
U = UAB + UBC
print(f'U = {UAB:.2f} + {UBC:.2f} = {U:.2f} J')
```

\begin{equation*}
U = P_\mathrm{anel} \, h 
\end{equation*}


```python
h = U/d.P
print(f'{h = :.4f} m')
```
