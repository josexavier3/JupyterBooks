### Problema 3

Cada barra da treliça indicada é feita de aço e tem a área da secção transversal uniforme e igual a 3200 mm $^2$. Considerando $E = 200$ GPa, determine o deslocamento vertical do nó C provocado pela aplicação da força $P= 66,7$ kN. Dados: $a = 0.75 $ m e $b = 1,8 $ m.

## Resolução

O deslocamento de um ponto da estrutura onde é aplicada uma única força pontual pode ser calculado em função da energia de deformação

\begin{equation*}
U = \frac{1}{2}P\delta_C
\quad\Leftrightarrow\quad
\delta_C = \frac{2U}{P}
\end{equation*}

onde a energia de deformação pode ser calculada pela expressão,

\begin{equation*}
U = \sum_i \frac{F_i²L_i}{2A_iE_i}
\end{equation*}

Ou seja, para a determinação do deslocamento $\delta_C$ deve-se conhecer os esforços em todas as barras da estrutura. Estes esforços podem ser determinados pelo método dos nós.

#### Determinação das forças nas barras pelo método dos nós

Análise do equilíbrio da treliça, pelo método dos nós, para a determinação dos esforços axiais suportados por cada barra.

A aplicação do método dos nós implica a escolha *a priori* da sequência de nós. Deve-se sempre iniciar com um nó com apenas 2 incógnitas em termos de forças nas barras. Os nós seguintes devem seguir este critério uma vez que só se podem aplicar 2 equações de equilíbrio estático: $\sum F_x = 0; \sum F_y = 0$.

Uma pré-análise da forma como as barras se ligam aos nós (número e orientação), permite ainda concluir que barras terão um esforço nulo.

##### Pré-análise do nó B:

De notar que nó $B$ ligam-se 3 barras. Sendo que a barra AB e BC são colineares e uma barra BD, perpendicular a estas. Desta forma, é possivel concluir:

\begin{equation*}
|F_{AB}| = |F_{BC}|~\textrm{(sentidos opostos)} \quad\wedge\quad F_{BD} = 0
\end{equation*}


```python
FBD = 0.; print(f'{FBD = } [kN]')
```

##### Pré-análise do nó E:


De notar que nṍ $E$ ligam-se 2 barras e existe uma ligação ao exterior por um apoio simples. A força na barra ED irá equilibrar a reação em E, pelo que a força da barra AE é nula (barra de esforço nulo):

\begin{equation*}
|F_{ED}| = |R_{Ex}|~\textrm{(sentidos opostos)} \quad\wedge\quad F_{AE} = 0
\end{equation*}

```python
FAE = 0.; print(f'{FAE = } [kN]')
```

##### Sequência de nós usada no método dos nós

Podemos usar a seguinte sequência de nós para o método dos nós: C $\rightarrow$ D. De notar que basta a análise do equilíbro destes dois nós para este problema.

- Nó C:

Assumindo que as barras estão em tração por hipótese, o diagrama de corpo livre do nó C ilustra 3 forças: a força vertical diretamente aplicada; força horizonta da barra BC e a força inclinada da barra DC:

\begin{equation*}
\begin{cases}
\rightarrow(+)&:& \sum F_x = 0 \\
\uparrow(+) &:&\sum F_y = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
-F_{BC} - F_{DC}\cos\alpha = 0 \\
-P - F_{DC}\sin\alpha = 0
\end{cases}
\quad\wedge\quad
\alpha = \arctan\left(\frac{W}{H}\right)
\end{equation*}

```python
fbc, fdc = sy.symbols('fbc fdc')
ang = np.arctan(d.H/d.W)
print(f'ang = {ang:.3f} [rad] =  {np.rad2deg(ang):.2f} [deg]')
sumFX = -fbc - fdc*np.cos(ang); print(f'sumFX = {sumFX}')
sumFY = -d.P - fdc*np.sin(ang); print(f'sumFY = {sumFY}')
sol = solve([sumFX,sumFY],[fbc, fdc])
FBC = sol[fbc]; print(f'FBC = {FBC*1e-3:.1f} [kN] | {np.abs(FBC*1e-3):.1f} (T) [kN] ')
FDC = sol[fdc]; print(f'FDC = {FDC*1e-3:.1f} [kN] | {np.abs(FDC*1e-3):.1f} (C) [kN] ')
FAB = FBC; print(f'FAB = {FAB*1e-3:.1f} [kN] | {np.abs(FAB*1e-3):.1f} (T) [kN] ')
```

- Nó D:

Assumindo que as barras estão em tração por hipótese, o diagrama de corpo livre do nó D ilustra 3 forças: a força da barra DC, a força da barra AD e a força da barra ED:

\begin{equation*}
\begin{cases}
\rightarrow(+)&:& \sum F_x = 0 \\
\uparrow(+) &:&\sum F_y = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
-F_{ED} - F_{AD}\cos\alpha + F_{DC}\cos\alpha = 0 \\
F_{AD}\sin\alpha + F_{DC}\sin\alpha = 0
\end{cases}
\quad\wedge\quad
\alpha = \arctan\left(\frac{W}{H}\right)
\end{equation*}

```python
fed, fad = sy.symbols('fed fad')
sumFX = -fed - fad*np.cos(ang) + FDC*np.cos(ang); print(f'sumFX = {sumFX}')
sumFY = fad*np.sin(ang) + FDC*np.sin(ang); print(f'sumFY = {sumFY}')
sol = solve([sumFX,sumFY],[fed, fad])
FED = sol[fed]; print(f'FED = {FED*1e-3:.1f} [kN] | {np.abs(FED*1e-3):.1f} (C) [kN] ')
FAD = sol[fad]; print(f'FAD = {FAD*1e-3:.1f} [kN] | {np.abs(FAD*1e-3):.1f} (T) [kN] ')
```

- Cálculo das reações

Embora não seja pedido, pode-se determinar as reações nos apoios, e confirmar a relação: $F_{ED} = R_{Ex}$

Construção do diagrama de corpo livre e análise do equilíbrio estático:

\begin{equation*}
\sum F_x = 0
\quad\wedge\quad
\sum F_y = 0
\quad\wedge\quad
\sum M_A = 0
\end{equation*}

Resultando no sistema de equações:

\begin{equation*}
\begin{cases}
R_{Ax} + R_{Ex} & = 0 \\
R_{Ay} - P & = 0 \\
-R_{Ex} L_{AE} - PL_{AC} &= 0
\end{cases}
\end{equation*}


```python
rax, ray, rex  = sy.symbols('rax ray rex')

print('::: sistema de equações :::')
sumFX = sy.Eq(rax + rex,0); print(f'{sumFX = } [N]')
sumFY = sy.Eq(ray - d.P,0); print(f'{sumFY = } [N]')
sumMA= sy.Eq(rex*d.H - d.P*(2*d.W),0)
print(f'{sumMA = } [N.m]')
sol = solve([sumFX,sumFY,sumMA],[rax, ray, rex])
print(f'{sol = }')
```

```python
print('::: resolução :::')
RAx = sol[rax]; print(f'RAx = {RAx*1e-3:.1f} [kN]')
RAy = sol[ray]; print(f'RAy = {RAy*1e-3:.1f} [kN]')
REx = sol[rex]; print(f'REx = {REx*1e-3:.1f} [kN]')
```

```python
print('| Barra i | Força, Fi [N] | Comprimento, Li [m] | Fi².Li [Nm] |')
print('---------------------------------------------------------------')

# Define a list of tuples containing the bar names, forces, and lengths
bars = [
    ('AB', FAB, d.W),
    ('AD', FAD, np.sqrt(d.W**2 + d.H**2)),
    ('AE', FAE, d.H),
    ('ED', FED, d.W),
    ('BD', FBD, d.H),
    ('BC', FBC, d.W),
    ('DC', FDC, d.W)
]

# Initialize the sum of Fi².Li
sum_F_L = 0

# Loop through each bar and calculate Fi².Li
for name, force, length in bars:
    F_L = force**2 * length
    sum_F_L += F_L
    print(f'| {name:<6}  | {force:>7.1f}      | {length:.2f}                | {F_L:.3e}   |')

print('---------------------------------------------------------------')
print(f' SUM :                                            {sum_F_L:.3e} ')
```

#### Energia de deformação

A energia de deformação pode ser calculada pela expressão,

\begin{equation*}
U = \sum_i \frac{F_i²L_i}{2A_iE_i}
\end{equation*}

Uma vez que as barras são constituidas do mesmo material ($E$) e secção transversal,

\begin{equation*}
U = \frac{1}{2AE}\sum F_i²L_i
\end{equation*}

```python
U = sum_F_L/2/d.E/d.A
print(f'{U = :.1f} [J]')
```

Uma vez que existe uma única carga aplicada na estrutura, o deslocamento do ponto de aplicação da força $P$ pode ser determinado por,

\begin{equation*}
W = U = \frac{1}{2}P\delta_C
\quad\Leftrightarrow\quad
\delta_C = \frac{2U}{P}
\end{equation*}

```python
deltaC = (2*U)/d.P
print(f'deltaC = {deltaC:.3e} [m] = {deltaC*1e3:.2f} [mm]')
```

Salienta-se que este método de igualar a energia de deformação ao trabalho da força aplicada pode ser usado porque existe apenas uma única carga aplicada.

Genericamente, deve-se usar o Teorema de Castigliano para a determinação de deslocamentos em pontos da estrutura, e este pode ser aplicado quando existem vários carregamentos aplicados sobre a estrutura.
