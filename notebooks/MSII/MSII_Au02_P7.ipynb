### Problema 7

Considere a secção transversal da viga de betão armado representada na figura que é reforçada com quatro varões de aço com diâmetro $d =$ 25 mm. Sabendo que a viga suporta um momento flector positivo $M = $ 180 kN/m e que o módulo de elasticidade do betão é $E_b =$ 25 GPa e o do aço é $E_a$ = 200 GPa, determine as tensões máximas que são geradas no betão e no aço.

```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt

cor = '2'
if cor == '1':
    plt.rcParams['axes.facecolor'] = (.15, .15, .15)
    plt.rcParams['figure.facecolor'] = (.15, .15, .15)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.8, .8, .8),
              "xtick.color" : (.8, .8, .8),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.7, .7, .7),
              "axes.labelcolor" : (.8, .8, .8),
              "axes.edgecolor" : (.15, .15, .15)}
else:
    plt.rcParams['axes.facecolor'] = (.7, .7, .7)
    plt.rcParams['figure.facecolor'] = (.7, .7, .7)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.1, .1, .1),
              "xtick.color" : (.1, .1, .1),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.1, .1, .1),
              "axes.labelcolor" : (.1, .1, .1),
              "axes.edgecolor" : (.15, .15, .15)}
plt.rcParams.update(params)

# Define data structures with units in N, mm, and MPa
class Material:
    def __init__(self, E):
        self.E = E  # Young's Modulus in MPa

class Section:
    def __init__(self, l1, l2, h1, h2, h3, d):
        self.l1 = l1  # mm
        self.l2 = l2  # mm
        self.h1 = h1  # mm
        self.h2 = h2  # mm
        self.h3 = h3  # mm
        self.ht = h1 + h2 + h3  # Total height in mm
        self.d = d
    
    @property # The @property decorator is used to define a method that can be accessed like an attribute
    def Area(self):
        return 4*np.pi*self.d**2/4 # 4 times de diameter       

# Define constants
Mf = 180.1e6  # Bending moment in N.mm

# Create instances of material and section
Aco = Material(E=200e3)  # Steel material with Young's Modulus
Bet = Material(E=25e3)   # Another material with Young's Modulus

S = Section(l1=300, l2=100, h1=60, h2=420, h3=120, d=25)  # Section dimensions
print(f' Area inicial aço = {S.Area:.1f} mm2')
```

```python
n = Aco.E / Bet.E  
print(f'n = {n:.1f}')
print(f'area do aco inicial = {S.Area:.2f} (mm2)')
S.AreaAco2Betao = S.Area * n
print(f'area equivalente do aco tranformado em betao = {S.AreaAco2Betao:.2f} (mm2)')
lAco2Betao = S.AreaAco2Betao / S.d
print(f'l equivalente do aco tranformado em betao = {lAco2Betao:.2f} (mm)')
```

```python
k = sy.symbols('k')
EQ = (k-S.h1)*(2*S.l1*S.h3) + (k-S.h3)*(S.l1)*(k-S.h3)/2 - (S.h2 + S.h3 - k)*S.AreaAco2Betao
simplified_EQ = sy.simplify(EQ)
print(f'Eq(k) = {sy.simplify(EQ)}')
sol = sy.solve(EQ,k)
K = sol[1]
print(f'K = {K:.1f} mm')
```

```python
# Calculo de It
def iret(b,h): return b*h**3/12

print('::::::::::::::::::::::::::::::')
A1 = S.l1*S.h3
print(f'A1 = {A1:.1f} mm²')
d1 = K - S.h3/2
print(f'd1 = {d1:.1f} mm')
Ic1 = iret(S.l1,S.h3)
print(f'Ic1 = {Ic1:.1f} mm³')
I1 = 2*(Ic1 + A1*d1**2)
print(f'I1 = {I1:.1f} mm³')

print('::::::::::::::::::::::::::::::')
A2 = S.l1*(K-S.h3)
print(f'A2 = {A2:.1f} mm²')
d2 = (K-S.h3)/2
print(f'd2 = {d2:.1f} mm')
Ic2 = iret(S.l1,K-S.h3)
print(f'Ic2 = {Ic2:.1f} mm³')
I2 = Ic2 + A2*d2**2
print(f'I2 = {I2:.1f} mm³')

print('::::::::::::::::::::::::::::::')
A3 = S.AreaAco2Betao
print(f'A3 = {A3:.1f} mm²')
d3 = (S.h3+S.h2)-K
print(f'd3 = {d3:.1f} mm')
Ic3 = iret(lAco2Betao,S.d)
print(f'Ic3 = {Ic3:.1f} mm³')
I3 = Ic3 + A3*d3**2
print(f'I3 = {I3:.1f} mm³')

print('\n::::::::::::::::::::::::::::::')
Iz = I1 + I2 + I3
print(f'Iz = {Iz:.1f} mm³')
```

```python
# Calculo sigma_Betao_max e sigma_Aco_max
yb = K
Sb = -Mf/Iz*yb
print(f'Sb = {Sb:.1f} MPa')

ya = - (S.h3+S.h2)-K
Sa = -Mf/Iz*ya*n
print(f'Sa = {Sa:.1f} MPa')
```
