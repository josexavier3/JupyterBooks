### Resolução

Nos problemas do capítulo anterior, analisou-se o dimensionamento de vigas, considerando a tensão normal atuante na secção transversal da viga devido ao momento fletor. Neste capítulo são abordadas as tensões de corte resultantes do esforço transversal na viga, o que é especialmente importante na análise de vigas curtas.

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
import matplotlib.pyplot as plt
import os

cor = '2'
if cor == '1':
    plt.rcParams['axes.facecolor'] = (.15, .15, .15)
    plt.rcParams['figure.facecolor'] = (.15, .15, .15)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.8, .8, .8),
              "xtick.color" : (.8, .8, .8),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.7, .7, .7),
              "axes.labelcolor" : (.8, .8, .8),
              "axes.edgecolor" : (.15, .15, .15)}
else:
    plt.rcParams['axes.facecolor'] = (.7, .7, .7)
    plt.rcParams['figure.facecolor'] = (.7, .7, .7)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.1, .1, .1),
              "xtick.color" : (.1, .1, .1),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.1, .1, .1),
              "axes.labelcolor" : (.1, .1, .1),
              "axes.edgecolor" : (.15, .15, .15)}
plt.rcParams.update(params)

# data structure, units: N, mm, MPa
# Create an empty class
class varin: pass

d = varin()
d.L = 0.150 # m
d.L1 = 0.020
d.W1 = 0.020
d.W2 = 0.150
d.h = 0.090
d.pregFmax = 200. # N
d.pregfilas = 2.

d.W = 2*d.W1 + d.W2
d.A = d.W*d.L
print(f'A = {d.L:.3f} x {d.W:.3f} = {d.A:.3e} m²')
```

Considere uma ligação rígida entre os blocos de madeira, assegurada pela ligação aparafusada. Os pregos são dispostos em intervalos de 90 mm, em duas filas, tanto na parte superior como na parte inferior das tábuas.

#### Fluxo de corte:

Entre o bloco superior e as tabuas verticais é possível calcular um fluxo de corte sobre as superfícies de contato, dado por:

\begin{equation*}
q = \frac{V Q}{I_z} \quad\textrm{[N/m]}
\end{equation*}

onde:

- $I_z$ : 2º momento de área ou momento de inércia de toda a secção transversal.

- $Q$ : 1º momento de área da secção em corte (esta secção resistente ao corte deve ser destacada e isolada; neste caso ela corresponde à superfície de contacto entre as tábuas verticais e horizontais).

- $V$ : Esforço de corte resultante aplicado na secção longitudinal em análise.

A equação anterior pode ser resolvida para o esforço transverso máximo ($V$), atendendo que a força de corte que cada prego suporta é de $F_\mathrm{max/prego}=200$ N:

\begin{equation*}
V = q\frac{I_z}{Q} \quad\textrm{[Pa]}
\end{equation*}

Para a resolução do problema deve-se neste caso determinar: $q$, $I_z$ e $Q$.


#### $Q$ : 1º momento de área da secção em corte

O Eixo Neutro é completamente definido pelos eixos de simetria da secção. O primeiro momento de área deve ser calculado em relação a este referencial principal de inércia.

Para a determinação do valor de $Q$ deve-se isolar a tábua na secção onde se pretende avaliar a tensão de corte. Neste caso temos interesse em isolar a parte superior horizontal, escrevendo:

\begin{equation*}
Q = A_1 \overline{y}_1
\end{equation*}

onde:

- $A_1$ : é a área total da região isolado da tábua superior horizontal (150 $\times$ 20 mm $^2$)
- $\overline{y}_1$ : é a distância vertical entre a linha que contem o centróide da área isolada da tábua superior horizontal e o eixo neutro.

```python
y_1 = d.W1/2 + d.W2/2
print(f'y_1 = {y_1:.4f} [m]')
A_1 = d.L*d.W1
print(f'A_1 = {A_1:.4f} [m²]')
Q = A_1*y_1
print(f'Q = {Q:.3e} [m³]')
```

#### $I_z$ : 2º momento de área da secção em corte

Uma vez que a secção é homogénea, determina-se o momento de inércia de área (total) por operações booleanas e o formulário de uma secção elementar retangular.

```python
def iz(b,h): return b*h**3/12

Iz_rectbig = iz(d.L,d.W)
print(f'Iz (rect big) = {Iz_rectbig:.3e} m⁴')
Iz_rectsmall = iz(d.L-2*d.L1,d.W2)
print(f'Iz (rect small) = {Iz_rectsmall:.3e} m⁴')
Iz = Iz_rectbig - Iz_rectsmall
print(f'Iz = {Iz:.3e} m⁴')
```

#### Ligação pregada em madeira

O fluxo de forte ($q$) ou tensão de corte por unidade de comprimento ao longo da ligação pregada deve ser igualmente distribuída pelo número total de pregos que suportam a ligação.

A ligação pregada é realizada pela fixação de $N$ pregos, em duas filas, em cada tábua superior e inferior.

Quantos pregos ($N$) por metro de comprimento (1 m) estão presentes nesta ligação, sabendo que estes distam entre si de 90 mm?

Pode-se usar o seguinte formulário:

\begin{equation*}
h = \frac{F_\mathrm{max}}{q}, 
~\quad~\textrm{o esforço de corte é cntudo distribuido por $m$ filas}
\quad\therefore\quad
m F_\mathrm{max} = q h
\end{equation*}

em que,

- $h$ : espaçamento entre pregos ao longo da viga [m]
- $m$ : número de filas ou repetições de parafusos que atravessam a área usada para o cálculo de $Q$
- $F_\mathrm{max}$ : capacidade de carga por unidade ou prego

Esta equação pode ainda escrever-se:

\begin{equation*}
q \left[\frac{\mathrm{N}}{\mathrm{m}}\right] = m N \left[\frac{\mathrm{unid. pregos}}{\mathrm{m}}\right] F_\mathrm{max} \left[\frac{\mathrm{N}}{\mathrm{unid. pregos}}\right]
\quad\wedge\quad N = \frac{1}{h}
\end{equation*}

onde,

- $m$ : número de filas ou repetições de parafusos ao longo do comprimento
- $N$ : número de pregos ao longo do comprimento de uma fila
- $F_\mathrm{max}$ : neste caso é o valor máximo que cada prego suporta


A ligação pregada é realizada pela fixação de $N$ pregos, em duas filas, em cada tábua superior e inferior.

Quantos pregos ($N$) por metro de comprimento (1 m) estão presentes nesta ligação, sabendo que estes distam entre si de 90 mm?

\begin{equation*}
\begin{aligned}
1 ~\mathrm{[m]} & \longrightarrow N ~\mathrm{[unid. pregos]}\\
0.090 ~\mathrm{[m]} & \longrightarrow 1 ~\mathrm{[unid. pregos]}\\
\end{aligned}
\end{equation*}

\begin{equation*}
\frac{N}{1} = \frac{1}{0.090}
\end{equation*}

```python
N = 1/d.h
print(f'N.º de pregos por metro em cada fila de preços = {N:.1f} pregos ~ {int(N)} pregos')
```

(Nota: na prática poder-se-ia colocar 11 pregos distantes entre si de aproximadamente 90 mm.)


```python

m = d.pregfilas
q = m*int(N)*d.pregFmax
print(f'm = {m:.0f} | N = {int(N)}| Fmax = {d.pregFmax} | q = {q:.1f} [N/m]')
```

Pelo que o esforço de corte máximo aplicável pelo critério de resistância da ligação:

\begin{equation*}
V = \frac{I_z }{Q} (m N F_\mathrm{max}) \quad\textrm{[N]}
\end{equation*}

```python
v = sy.symbols('v')
V = solve(q - v*Q/Iz,v)[0]
print(f'V = {V:.1f} [N]')
V_cal2 = Iz * m * int(N) * d.pregFmax / Q
print(f'V (cal2) = {V_cal2:.1f} [N]')

V_cal3 = Iz/Q*m*d.pregFmax/d.h
print(f'V (cal3) - usando o espaçamento h - = {V_cal3:.1f} [N]')
```
