## Resolução

- Cálculo das reações

Diagrama de corpo livre + análise de equilíbrio:

\begin{equation*}
\sum F_x = 0
\quad\wedge\quad
\sum F_y = 0
\quad\wedge\quad
\sum M_A = 0
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

# units: m, N
LAB: float = 0.6
LBC: float = 1.2
LBD: float = 2.0
dBD: float = 6.e-3
dAC: float = 50.e-3
P: float = 1500 # unit: N

rax, ray, rdy, p = sy.symbols('rax ray rdy p')

print('::: sistema de equações :::')
sumFX = rax
print(f'{sumFX = } [N]')
sumFY = ray + rdy - p
print(f'{sumFY = } [N]')
sumMA = rdy*LAB - p*(LAB+LBC)
print(f'{sumMA = } [N.m]')
```

```python
print('::: resolução :::')
sol = solve([sumFX,sumFY,sumMA],[rax, ray, rdy])
RAx = sol[rax]; print(f'{RAx = } [N]')
RAy = sol[ray]; print(f'{RAy = } = {RAy.subs(p,P)*1e-3:.1f} [kN]')
RDy = sol[rdy]; print(f'{RDy = } = {RDy.subs(p,P)*1e-3:.1f} [kN]')
```

A energia de deformação armazenada num estrutura com elementos à flexão e axial, submetido a um carregamento externo é expressa por,

\begin{equation*}
U = \int_0^L \frac{M²}{2EI} \mathrm{d}x + \int_0^{L'} \frac{N²}{2AE} \mathrm{d}x
\end{equation*}

Derivando a enerdia de deformação em relação à carga $P$, obtêm-se uma expressão para o cálculo do deslocamento no ponto e direção da força aplicada,

\begin{equation*}
\delta = \frac{\mathrm{d}U}{\mathrm{d}P} = \int_0^L \frac{M}{EI} \frac{\mathrm{d} M}{\mathrm{d} P} \mathrm{d}x
+ \int_0^{L'} \frac{N}{AE}\frac{\mathrm{d} N}{\mathrm{d} P} \mathrm{d}x
\end{equation*}

Para membros homogéneos,

\begin{equation*}
\boxed{
\delta = \frac{1}{EI} \int_0^L M \frac{\mathrm{d} M}{\mathrm{d} P} \mathrm{d}x +
\frac{1}{AE}\int_0^{L'} N\frac{\mathrm{d} N}{\mathrm{d} P} \mathrm{d}x
}
\end{equation*}

Assim sendo, a determinação do deslocamento implica a determinação do diagrama de momentos fletores ao longo da estrutura e sua derivada analitica em relação à carga $P$. Mais concretamente pode-se seguir o seguinte **algoritmo**:

1. Expressar a carga $P$ como uma variável: Se a carga externa atua na viga no ponto onde a deflexão deve ser calculada e na direção da linha de ação da força, em seguida, deve-se designar essa carga como a variável, e.g. $P$. Como resultado, os cálculos subsequentes serão realizado em termos da variável $P$, em vez do seu valor numérico real.
2. Determinar a expressão analítica do momento fletor interno $M$ em termos de $P$.
3. Derivadas parciais: derivar as expressões para o momento fletor interno em relação a $P$: $\mathrm{d} M /\mathrm{d} P$.
4. Substitua o valor numérico de $P$.
5.Integração: executar a integração para calcular a deflexão desejada. Uma resposta positiva indica que a deflexão atua na mesma direção
que $P$, e uma resposta negativa indica na direção oposta.

### Diagrama de esforços

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au10/P6/MSII_Au10_P6_i.png"
style="width: 30%;"/>

-  **Troço AB**

\begin{equation*}
x_1 \in [A, B [ ~: \qquad
\sum M_{AB} = 0 \Leftrightarrow M_{AB}(x_1) - R_{Ay}x_1= 0
\quad\wedge\quad
R_{Ay} = -2P
\end{equation*}

```python
x, ei = sy.symbols('x ei')
MAB = RAy*x
print(f'{MAB = } = {MAB.subs(p,P)} [N.m]')
dMABdP = sy.diff(MAB,p)
print(f'{dMABdP = } = {dMABdP.subs(p,P)} [N.m]')
dC1 = sy.integrate(MAB*dMABdP,(x,0,LAB))/ei
print(f'{dC1 = } = {dC1.subs(p,P)} [N.m]')
```

- **Troço CB**

\begin{equation*}
x_2 \in [C, B [ ~: \qquad
\sum M_{CB} = 0 \Leftrightarrow M(x_2) - Px_2= 0
\end{equation*}

```python
MCB = p*x
print(f'{MCB = } = {MCB.subs(p,P)} [N.m]')
dMCBdP = sy.diff(MCB,p)
print(f'{dMCBdP = } [N.m]')
dC2 = sy.integrate(MCB*dMCBdP,(x,0,LBC))/ei
print(f'{dC2 = } = {dC2.subs(p,P)} [m]')
```

- **Troço DB**

\begin{equation*}
x_3 \in [D, B [ ~: \qquad
\sum F_x = 0 \Leftrightarrow N(x_3) - R_{Dy} = 0
\quad\wedge\quad
R_{Dy} = 3 P
\end{equation*}

```python
ae = sy.symbols('ae')
NDB = RDy
print(f'{NDB = } = {NDB.subs(p,P):.1f} [N.m]')
dNDBdP = sy.diff(NDB,p)
print(f'{dNDBdP = :.1f} = {dNDBdP.subs(p,P):.1f} [N.m]')
dC3 = sy.integrate(NDB*dNDBdP,(x,0,LBD))/ae
print(f'{dC3 = } = {dC3.subs(p,P)} [m]')
```

- **Resultado final:**

\begin{equation*}
\delta_C = \left.\delta_C(M)\right|_{AB} + \left.\delta_C(M)\right|_{CB} + \left.\delta_C^{N}\right|_{DB} = \frac{\mathrm{d}U}{\mathrm{d}P}
\end{equation*}

desenvolvendo,

\begin{equation*}
\delta_C
= \int_A^B \frac{M_{AB}}{EI} \frac{\mathrm{d} M_{AB}}{\mathrm{d} P} \mathrm{d}x
+ \int_C^B \frac{M_{CB}}{EI} \frac{\mathrm{d} M_{CB}}{\mathrm{d} P} \mathrm{d}x
+ \int_D^B \frac{N_{DB}}{AE}\frac{\mathrm{d} N_{DB}}{\mathrm{d} P} \mathrm{d}x
\end{equation*}

```python
def acirc(r): return np.pi*r**2

Acabo = acirc(dBD/2)
print(f'{Acabo = :.3e} [m²]')
Avig = acirc(dAC/2)
print(f'{Avig = :.3e} [m²]')

def icirc(r): return np.pi/4*r**4

E: float = 200.e9 # unit Pa
I = icirc(dAC/2) # m⁴
print(f'{I = :.3e} [m²]')
EI = E*I
print(f'{EI = :.3e} [N.m²]')
AE = E*Acabo
print(f'{AE = :.3e} [N]')

dC = dC1 + dC2 + dC3
print(f'dC = {dC1} + {dC2} + {dC3} [m]')
dC1n, dC2n, dC3n = dC1.subs([(p,P),(ei,EI)]), dC2.subs([(p,P),(ei,EI)]), dC3.subs([(p,P),(ae,AE)])
dCn = dC1n + dC2n + dC3n
print(f'dC = {dC1n:.3e} + {dC2n:.3e} + {dC3n:.3e} = {dCn:.3e} [m] = {dCn*1e3:.2f} [mm] ')
```

## Teorema da Carga Unitária

O Teorema da Carga Unitária é um método para calcular deslocamentos (ou rotações) em estruturas usando o Princípio dos Trabalhos Virtuais. A ideia fundamental é:

**Para calcular o deslocamento num ponto e direção específicos:**
1. Aplica-se uma carga unitária fictícia (força = 1 ou momento = 1) no ponto e direção onde se pretende conhecer o deslocamento
2. Calcula-se o diagrama de momentos fletores devido a esta carga unitária: **Mu(x)**
3. Calcula-se o diagrama de momentos fletores devido ao carregamento real: **M(x)**
4. Integra-se o produto dos dois diagramas ao longo da estrutura

Como mostra a expressão na tua imagem:

$$1 \cdot \delta_D = \sum_i \int_0^{L_i} \frac{M(x)M_u(x)}{EI} \, dx$$

Onde:
- **δD** = deslocamento pretendido
- **M(x)** = momento fletor devido ao carregamento real
- **Mu(x)** = momento fletor devido à carga unitária
- **EI** = rigidez à flexão

## Teorema de Castigliano

O **Segundo Teorema de Castigliano** (o mais usado) estabelece que:

> *A derivada parcial da energia de deformação em relação a uma força (ou momento) aplicada é igual ao deslocamento (ou rotação) no ponto de aplicação dessa força, na sua direção.*

Matematicamente:

$$\delta_i = \frac{\partial U}{\partial P_i}$$

Onde:
- **U** = energia de deformação total da estrutura
- **Pi** = força aplicada no ponto i
- **δi** = deslocamento no ponto i na direção de Pi

Para flexão, a energia de deformação é:

$$U = \sum_i \int_0^{L_i} \frac{M(x)^2}{2EI} \, dx$$

## Relação entre os Teoremas

**Os dois teoremas são matematicamente equivalentes!** A demonstração vem de:

1. Pelo Teorema de Castigliano:
$$\delta = \frac{\partial U}{\partial P} = \frac{\partial}{\partial P} \left[\int \frac{M(x)^2}{2EI} dx\right]$$

2. Aplicando a derivada (e lembrando que M depende de P):
$$\delta = \int \frac{M(x)}{EI} \cdot \frac{\partial M(x)}{\partial P} \, dx$$

3. Mas $\frac{\partial M(x)}{\partial P}$ é exatamente o momento fletor causado por uma carga unitária aplicada em P, ou seja, **Mu(x)**!

4. Logo:
$$\delta = \int \frac{M(x) \cdot M_u(x)}{EI} \, dx$$

Que é precisamente a expressão do Teorema da Carga Unitária.

## Diferenças Práticas

| **Aspecto** | **Castigliano** | **Carga Unitária** |
|-------------|-------------------|-----------------|
| **Quando usar** | Quando já existe uma força no ponto de interesse | Quando não há força no ponto (adiciona-se uma fictícia) |
| **Procedimento** | Derivação da energia | Dois estados: real + unitário |
| **Vantagem** | Conceitualmente mais simples | Pode ser mais direto para estruturas complexas |
