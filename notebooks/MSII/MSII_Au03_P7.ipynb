```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
import matplotlib.pyplot as plt
import os

cor = '2'
if cor == '1':
    plt.rcParams['axes.facecolor'] = (.15, .15, .15)
    plt.rcParams['figure.facecolor'] = (.15, .15, .15)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.8, .8, .8),
              "xtick.color" : (.8, .8, .8),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.7, .7, .7),
              "axes.labelcolor" : (.8, .8, .8),
              "axes.edgecolor" : (.15, .15, .15)}
else:
    plt.rcParams['axes.facecolor'] = (.7, .7, .7)
    plt.rcParams['figure.facecolor'] = (.7, .7, .7)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.1, .1, .1),
              "xtick.color" : (.1, .1, .1),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.1, .1, .1),
              "axes.labelcolor" : (.1, .1, .1),
              "axes.edgecolor" : (.15, .15, .15)}
plt.rcParams.update(params)

# data structure, units: N, mm, MPa
# Create an empty classclass varin: pass

from dataclasses import dataclass

@dataclass
class Varin:
    F1: float
    F2: float
    F3: float
    b1: float
    b2: float
    tf: float
    tw: float
    d: float

inp = Varin(F1=30.e3, F2=30.e3, F3=50.e3, b1=0.160, b2=0.320, tf=0.020, tw=0.020, d=0.1)
alt_tot = 2*inp.tf + inp.d
print(vars(inp))
```

```python
print(f'Calculo do centroide da seccao')
A1, y1 = inp.b1*inp.tf, inp.tf + inp.d + inp.tf/2
A2, y2 = inp.d*inp.tw, inp.tf + inp.d/2
A3, y3 = inp.b2*inp.tf, inp.tf/2
A_tot = A1 + A2 + A3
y_cent = (A1*y1 + A2*y2 + A3*y3)/A_tot
print(f'y = {A1}({y1}) + {A2}({y2}) + {A3}({y3})/ {A_tot} = {y_cent:.4f} mm')
```

```python
N = inp.F1 + inp.F2 + inp.F3
print(f'N = {N:.2f} N')
Mz = -inp.F1*(alt_tot-y_cent) + inp.F2*y_cent + inp.F3*y_cent
print(f'Mz = -{inp.F1}({alt_tot - y_cent:.4f}) + {inp.F2+inp.F3}({y_cent:.4f}) = {Mz:.2f} N.mm')
My = -inp.F1*inp.b1/2 + inp.F2*inp.b2/2 - inp.F3*inp.b2/2
print(f'My = -{inp.F1}({inp.b1/2:.2f}) + ({inp.F2-inp.F3})({inp.b2/2:.2f}) = {My:.2f} N.mm')
```

```python
def Irect(b, h):
    return b*h**3/12
d1 = (alt_tot - y_cent) - inp.tf/2
Iz1 = Irect(inp.b1, inp.tf) + A1*d1**2
print(f'Iz1 = {Irect(inp.b1, inp.tf):.4e} + {A1}({d1:.4f}**2) = {Iz1:.4e} mm^4')

d2 = alt_tot/2 - y_cent
Iz2 = Irect(inp.tw, inp.d) + A2*d2**2
print(f'Iz2 = {Irect(inp.tw, inp.d):.4e} + {A2}({d2:.4f}**2) = {Iz2:.4e} mm^4')

d3 = y_cent - inp.tf/2
Iz3 = Irect(inp.b2, inp.tf) + A3*d3**2
print(f'Iz3 = {Irect(inp.b2, inp.tf):.4e} + {A3}({d3:.4f}**2) = {Iz3:.4e} mm^4')

Iz  = Iz1 + Iz2 + Iz3
print(f'Iz = {Iz:.4e} mm^4')
```

```python
Iy1 = Irect(inp.tf, inp.b1)
print(f'Iy1 = {Irect(inp.tf, inp.b1):.4e} mm^4')
Iy2 = Irect(inp.d, inp.tw)
print(f'Iy2 = {Irect(inp.d, inp.tw):.4e} mm^4')
Iy3 = Irect(inp.tf, inp.b2)
print(f'Iy3 = {Irect(inp.tf, inp.b2):.4e} mm^4')

Iy = Iy1 + Iy2 + Iy3
print(f'Iy  = {Iy:.4e} mm^4')
```

```python
import sympy as sy
def printsym(expr):
    return expr.xreplace({n : round(n, 3) for n in expr.atoms(sy.Number)})

snorm = N/A_tot
y, z = sy.symbols('y z')

if N > 0:
    EQ = snorm - Mz/Iz*y + My/Iy*z
else:
    EQ = snorm + Mz/Iz*y - My/Iy*z
print(f' Sxx = {printsym(EQ*1e-6)} [MPa]')
```

```python
print(f'Sxx = {printsym(EQ)} [Pa]')
print(f'    = {printsym(EQ*1e-6)} [MPa]')
sol = solve(EQ, y)
yfun = sol[0]
print(f'  y = {printsym(yfun)} [m]')
print(f'    = {printsym(yfun*1e3)} [mm]')
```

```python
def evalsxx(i,j):
    return EQ.subs([(z,i),(y,j)])*1e-6 # MPa

def print_stress(point, label):
    Sxx = evalsxx(point[0], point[1])
    print(f'Point {label}({point[0]:.3f},{point[1]:.3f}) mm : sigxx = {Sxx:.2f} MPa')

A = (inp.b2 / 2, -y_cent + inp.tf)
B = (-inp.b2 / 2, -y_cent)
C = (inp.b1 / 2, alt_tot - y_cent)

print_stress(A, 'A')
print_stress(B, 'B')
print_stress(C, 'C')
```
