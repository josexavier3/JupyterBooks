### Problema 7

A figura representa uma estrutura constituída pelas vigas AB, BC e CD construidas com um perfil HEB100 ( momento de 2ª ordem $I = 4,5\times 10^{-6}$ m $^4$ e área $2,6\times 10^{-3}$ m $^2$). O conjunto está ligado a um apoio fixo em A e, através do cabo DE, a um encastramento em E. O carregamento aplicado consiste numa carga distribuída uniforme aplicada entre B e C, $Q_0 = 30$ kN/m. Considere que o material das vigas e do cabo têm módulo de elasticidade $E = 210$ GPa e que a área da secção do cabo é $3,14\times 10^{-6}$ m $^2$. Determine, utilizando o teorema de Castigliano, o deslocamento horizontal no ponto D.

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
import matplotlib.pyplot as plt
import os

cor = '1'
if cor == '1':
    plt.rcParams['axes.facecolor'] = (.15, .15, .15)
    plt.rcParams['figure.facecolor'] = (.15, .15, .15)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.8, .8, .8),
              "xtick.color" : (.8, .8, .8),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.7, .7, .7),
              "axes.labelcolor" : (.8, .8, .8),
              "axes.edgecolor" : (.15, .15, .15)}
else:
    plt.rcParams['axes.facecolor'] = (.7, .7, .7)
    plt.rcParams['figure.facecolor'] = (.7, .7, .7)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.1, .1, .1),
              "xtick.color" : (.1, .1, .1),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.1, .1, .1),
              "axes.labelcolor" : (.1, .1, .1),
              "axes.edgecolor" : (.15, .15, .15)}
plt.rcParams.update(params)

# data structure, units: N, mm, MPa

class varin: pass

d = varin()
HEB100 = varin()

d.AB = 2.4  # unit: m
d.BC = 1.2  # unit: m
d.CD = 2.4  # unit: m
d.DE = 3.6  # unit: m

HEB100.I = 4.5e-6  # unit: m4
HEB100.A = 2.6e-3  # unit: m2
d.Q0 = 30.e3  # unit: N/m
d.E = 210.e9  # unit: Pa
d.Acabo = 3.14e-6 # unit: m2
```

## Formulário: teorema da carga unitária

\begin{equation*}
\boxed{W_\mathrm{tv} = U
\Leftrightarrow
1 \cdot \delta_{Du} = \sum_i \int_0^{L_i} \frac{M(x)M_u(x)}{EI} \mathrm{d}x}
\end{equation*}


## Resolução

- Diagrama de corpo livre

A estrutura está submetida ao carregamento linear vertical entre a barra BC. Além disso, no DCL são consideradas as reações sobre a estrutura. Existem no total 3 reações: (i) $R_{Ax}$, $R_{Ay}$; (ii) $R_{Ey}$. De notar que, apesar do encastramento, o cabo só absorve forças ao longo do seu comprimento.

- Cálculo das reações

\begin{equation*}
\begin{cases}
\rightarrow(+)&:& \sum F_x = 0 \\
\uparrow(+) &:&\sum F_y = 0 \\
\circlearrowleft(+)&:& \sum M_A = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
R_{Ax} + Q_0L_{BC} = 0 \\
R_{Ay} + R_{Ey} = 0 \\
-Q_0L_{BC}(L_{BC}/2) + R_{Ey}(L_{AB}+L_{CD}) = 0
\end{cases}
\end{equation*}

```python
rax, ray, rey = sy.symbols('rax ray rey')

print('::: sistema de equações :::')
Qeq = d.Q0*d.BC
print(f'{Qeq = } [kN]')
Qycent = d.BC/2
print(f'{Qycent = } [m]')
sumFX = rax + Qeq
print(f'{sumFX = } [N]')
sumFY = ray + rey
print(f'{sumFY = } [N]')
sumMD= -Qeq*Qycent + rey*(d.AB+d.CD)
print(f'{sumMD = } [N.m]')
sol = solve([sumFX,sumFY,sumMD],[rax, ray, rey])
```

```python
print('::: resolução :::')
RAx = sol[rax]; print(f'RAx = {RAx*1e-3:.1f} [kN]')
RAy = sol[ray]; print(f'RAy = {RAy*1e-3:.1f} [kN]')
REy = sol[rey]; print(f'REy = {REy*1e-3:.1f} [kN]')
```

#### Cálculo das expressões dos esforos internos (momento fletor nas vigas e esforço axial no cabo DE) pelo método das secções

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au10/P7/MSII_Au10_P7_i.png"
style="width: 40%;"/>

Notar que as expressões a seguir consideram o sentido positivo das forças de reação de acordo com o sentido obtido pela resolução do sistema de equilíbrio do DCL

### Vigas

- Tramo AB (corte à direifa)

\begin{equation*}
R_{Ay} x_1 + M_{AB}(x_1) = 0
\end{equation*}

- Tramo BC (corte à direita, vertical)

\begin{equation*}
-R_{Ax} x_2 + R_{Ay}L_{AB} + Q_0x_2\left(\frac{x_2}{2}\right)+ M_{BC}(x_2) = 0
\end{equation*}

- Tramo CD (corte à esquerda de D para C)

\begin{equation*}
R_{Ey} x_3 - M_{CD}(x_3) = 0
\end{equation*}

### Cabo

- Tramo DE

\begin{equation*}
R_{Ey} - N_{DE}(x) = 0
\end{equation*}

```python
x1, x2, x3 = sy.symbols('x1 x2 x3')
MAB = -np.abs(RAy)*x1
print(f'MAB = {MAB} [N.m]')
MBC = np.abs(RAx)*x2 - np.abs(RAy)*d.AB - d.Q0/2*x2**2
print(f'{MBC = }[N.m] ')
MCD = REy*x3
print(f'{MCD = }[N.m] ')
NDE = REy
print(f'{NDE = :.1f}[N] ')
```

##### Método da Carga Unitária

Para determinar o deslocamento de um nó e numa determinada direção, deve-se considerar a estrutura submetida a um carregamento unitário no ponto e direção do deslocamento.

- Diagrama de corpo livre

O DCL da estrutura consiste numa força pontual aplicada em D horizontal e as respetivas reações nos apoios em A e E.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au10/P7/MSII_Au10_P7_ii.png"
style="width: 40%;"/>


```python
raxu, rayu, reyu = sy.symbols('raxu rayu reyu')
load_u = 1

print('::: sistema de equações (carregamento unitário) :::')
sumFX = raxu + load_u
print(f'{sumFX = } [N]')
sumFY = rayu + reyu
print(f'{sumFY = } [N]')
sumMD= -load_u*d.BC + reyu*(d.AB+d.CD)
print(f'{sumMD = } [N.m]')
sol = solve([sumFX,sumFY,sumMD],[raxu, rayu, reyu])
```

```python
print('::: resolução :(carregamento unitário) ::')
RAxu = sol[raxu]; print(f'{RAxu = :.2f} [N]')
RAyu = sol[rayu]; print(f'{RAyu = :.2f} [N]')
REyu = sol[reyu]; print(f'{REyu = :.2f} [N]')
```

- Cálculo das expressões dos esforos internos (momento fletor nas vigas e esforço axial no cabo DE) pelo método das secções para o carregamento unitário

Notar que as expressões a seguir consideram o sentido positivo das forças de reação de acordo com o sentido obtido pela resolução do sistema de equilíbrio do DCL

### Vigas

- Tramo AB (corte à direifa)

\begin{equation*}
R_{Ay} x_1 + Mu_{AB}(x_1) = 0
\end{equation*}

- Tramo BC (corte à direita, vertical)

\begin{equation*}
-R_{Ax} x_2 + R_{Ay}L_{AB} + Mu_{BC}(x_2) = 0
\end{equation*}

- Tramo CD (corte à esquerda)

\begin{equation*}
R_{Ey} x_3 - Mu_{CD}(x_3) = 0
\end{equation*}

### Cabo

- Tramo DE

\begin{equation*}
R_{Ey} - Nu_{DE}(x) = 0
\end{equation*}

```python
MABu = -np.abs(RAyu)*x1
print(f'{MABu = } [N.m]')
MBCu = np.abs(RAxu)*x2 - np.abs(RAyu)*d.AB
print(f'{MBCu = } [N.m] ')
MCDu = REyu*x3
print(f'{MCDu = } [N.m] ')
NDEu = REyu
print(f'{NDEu = :.2f} [N] ')
```

- Teorema da Carga Unitária

\begin{equation*}
1 \cdot \delta_{Du} = \sum_i \left.\int_0^{L_i} \frac{M(x)M_u(x)}{EI}\right|_i\mathrm{d}x
\end{equation*}

- **Troço AB:**

\begin{equation*}
x_1 \in [A,B] = [0, 2.4] ~:\quad
\delta_{Du_1} = \int_{x_A}^{x_B} \left.\frac{M(x)M_u(x)}{EI}\right|_1\mathrm{d}x =
\frac{1}{EI}\int_{0}^{2.4}(-4.5x_1)\times10^3(-0.25x_1) \mathrm{d}x_1
\end{equation*}

```python
EI = d.E*HEB100.I
print(f'EI = {EI} [N.m2]')
du1 = 1/EI*sy.integrate(MAB*MABu,(x1,0,d.AB))
print(f'du1 = int [ 1/EI({MAB*MABu}) ]dx \n    = {du1:.3e} [m]')
```

- **Troço BC:**

\begin{equation*}
x_2 \in [B,C] = [0, 1.2] ~:\quad
\delta_{Du_2} = \left.\int_{x_B}^{x_C} \frac{M(x)M_u(x)}{EI}\right|_2\mathrm{d}x =
\frac{1}{EI}\int_{0}^{1.2}(-15x_2^2 + 36.0x_2 - 10.8)\times10^3(x_2-0.6) \mathrm{d}x_2
\end{equation*}

```python
du2 = 1/EI*sy.integrate(MBC*MBCu,(x2,0,d.BC))
print(f'du2 = int [ 1/EI({MBC*MBCu}) ]dx \n    = {du2:.3e} [m]')
```

- **Troço CD:**

\begin{equation*}
x_3 \in [C,D] = [0, 2.4] ~:\quad
\delta_{Du_3} = \left.\int_{x_C}^{x_D} \frac{M(x)M_u(x)}{EI}\right|_3\mathrm{d}x =
\frac{1}{EI}\int_{0}^{2.4}(4.5x_3)\times10^3(0.25x_3) \mathrm{d}x_3
\end{equation*}

```python
du3 = 1/EI*sy.integrate(MCD*MCDu,(x3,0,d.CD))
print(f'du3 = int [ 1/EI({MCD*MCDu}) ]dx \n    = {du3:.3e} [m]')
```

- **Troço DE:**

\begin{equation*}
x_4 \in [E,D] = [0, 3.6] ~:\quad
\delta_{Du_3} = \left.\int_{x_C}^{x_D} \frac{M(x)M_u(x)}{EI}\right|_3\mathrm{d}x =
\frac{1}{EI}\int_{0}^{3.6}(4.5)\times10^3(0.25) \mathrm{d}x_4
\end{equation*}

```python
x4 = sy.symbols('x4')
AE = d.E*d.Acabo
print(f'AE = {AE:.3e} [N.m⁴]')
du4 = 1/AE*sy.integrate(NDE*NDEu,(x4,0,d.DE))
print(f'du4 = int [ 1/EI({NDE*NDEu}) ]dx \n    = {du4:.3e} [m]')
```

- **Resultado final:**

\begin{equation*}
\delta_{Du} = \delta_{Du_1} + \delta_{Du_2} + \delta_{Du_3} + \delta_{Du_4}
\end{equation*}

```python
du = du1 + du2 + du3 + du4
print(f'du = {du1:.3e} + {du2:.3e} + {du3:.3e} + {du4:.3e} = {du:.3e} [m]')
```

## Teorema da Carga Unitária

O Teorema da Carga Unitária é um método para calcular deslocamentos (ou rotações) em estruturas usando o Princípio dos Trabalhos Virtuais. A ideia fundamental é:

**Para calcular o deslocamento num ponto e direção específicos:**
1. Aplica-se uma carga unitária fictícia (força = 1 ou momento = 1) no ponto e direção onde se pretende conhecer o deslocamento
2. Calcula-se o diagrama de momentos fletores devido a esta carga unitária: **Mu(x)**
3. Calcula-se o diagrama de momentos fletores devido ao carregamento real: **M(x)**
4. Integra-se o produto dos dois diagramas ao longo da estrutura

Como mostra a expressão na tua imagem:

$$1 \cdot \delta_D = \sum_i \int_0^{L_i} \frac{M(x)M_u(x)}{EI} \, dx$$

Onde:
- **δD** = deslocamento pretendido
- **M(x)** = momento fletor devido ao carregamento real
- **Mu(x)** = momento fletor devido à carga unitária
- **EI** = rigidez à flexão

## Teorema de Castigliano

O **Segundo Teorema de Castigliano** (o mais usado) estabelece que:

> *A derivada parcial da energia de deformação em relação a uma força (ou momento) aplicada é igual ao deslocamento (ou rotação) no ponto de aplicação dessa força, na sua direção.*

Matematicamente:

$$\delta_i = \frac{\partial U}{\partial P_i}$$

Onde:
- **U** = energia de deformação total da estrutura
- **Pi** = força aplicada no ponto i
- **δi** = deslocamento no ponto i na direção de Pi

Para flexão, a energia de deformação é:

$$U = \sum_i \int_0^{L_i} \frac{M(x)^2}{2EI} \, dx$$

## Relação entre os Teoremas

**Os dois teoremas são matematicamente equivalentes!** A demonstração vem de:

1. Pelo Teorema de Castigliano:
$$\delta = \frac{\partial U}{\partial P} = \frac{\partial}{\partial P} \left[\int \frac{M(x)^2}{2EI} dx\right]$$

2. Aplicando a derivada (e lembrando que M depende de P):
$$\delta = \int \frac{M(x)}{EI} \cdot \frac{\partial M(x)}{\partial P} \, dx$$

3. Mas $\frac{\partial M(x)}{\partial P}$ é exatamente o momento fletor causado por uma carga unitária aplicada em P, ou seja, **Mu(x)**!

4. Logo:
$$\delta = \int \frac{M(x) \cdot M_u(x)}{EI} \, dx$$

Que é precisamente a expressão do Teorema da Carga Unitária.

## Diferenças Práticas

| **Aspecto** | **Castigliano** | **Carga Unitária** |
|-------------|-------------------|-----------------|
| **Quando usar** | Quando já existe uma força no ponto de interesse | Quando não há força no ponto (adiciona-se uma fictícia) |
| **Procedimento** | Derivação da energia | Dois estados: real + unitário |
| **Vantagem** | Conceitualmente mais simples | Pode ser mais direto para estruturas complexas |
