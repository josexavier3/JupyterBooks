#### a) Escreva as equações do esforço transverso e do momento fletor em toda a viga e trace os diagramas desses esforços.

Neste exemplo, o elemento estrutural suportar cargas transversais aplicadas em vários pontos do seu comprimento. As cargas aplicadas são quer forças concentradas quer cargas distribuídas. Neste cenário, as cargas aplicadas resultam em esforço transverso
(tensão de Corte) e momento fletor (tensão normal). Para a análise e dimensionamento da viga deve-se calcular a posição e valor do momento fletor máximo. A tensão normal revela-se como o principal critério de dimensionamento:

\begin{equation}
\sigma_x = -\frac{M_z y}{I_z}
\quad\wedge\quad
\sigma_m = \frac{|M_z| c}{I_z} = \frac{|M_z|}{W}
\end{equation}

#### b) Sabendo que a secção transversal foi escolhida considerando uma tensão admissível
$\sigma_\mathrm{adm}$ = 250 MPa, verifique se essa tensão é ultrapassada em algum ponto da viga.

```python
Mm = 40*1e3 # units: N.m
ymax = HEB120.bf/2*1e-3 # unit: m
Iz = HEB120.Iz*1e-12 # unit: m
sigmax = Mm*ymax/Iz*1e-6

print(f'    Mm = {Mm:.1f}   [N.m]')
print(f'  ymax = {ymax:.3f}     [m]')
print(f'    Iz = {Iz:.3e} [m⁴]')
print(f'sigmax = {sigmax:.2f}    [MPa]')
```

- A tensão normal máxima deve ser inferior à tensão admissível do
material utilizado. Este critério permite a obtenção do menor
módulo de resistência necessário.

\begin{equation}
\sigma_\mathrm{m} \leq \sigma_\mathrm{adm}
\quad\wedge\quad
W_\mathrm{min} = \frac{|M|_\mathrm{max}}{\sigma_\mathrm{adm}}
\end{equation}

Conclui-se portanto que na face do perfil da viga, o critério de Tresca é ultrapasso: $\sigma_\mathrm{I} - \sigma_\mathrm{III} \gt \sigma_\mathrm{adm}$

#### c) Considere agora que é possível alterar o valor da força aplicado em A e determine qual o valor de
$F_\mathrm{A}$ que minimiza o momento fletor máximo na viga.

- O momento máximo negativo é aplicado em B:

\begin{equation}
M_B = -2 F_A
\end{equation}

- O momento máximo positivo ocorre em C:

Pela análise do equilíbrio da viga no troço AC ...

\begin{equation}
M_C = 51.4287 - 1.14287 F_A
\end{equation}

- Equacionando os momentos em B e C:

\begin{equation}
\begin{split}
|M_B| = |M_C|
& \Leftrightarrow
-2 F_A = 51.4287 - 1.14287 F_A \\
& \therefore
F_A = 16.364~\textrm{kN}
\quad\wedge\quad
M_B = -32.727~\textrm{KN.m}
\quad\wedge\quad
M_C = 32.727~\textrm{KN.m}
\end{split}
\end{equation}
