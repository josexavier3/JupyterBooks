### Notas prévias

##### Teorema da Carga Unitária

O método da carga unitária proporciona uma forma expedita de calcular as derivadas dos esforços internos necessárias para aplicar o Teorema de Castigliano.

A derivada de um esforço é igual ao esforço provocado por uma carga unitária aplicada no ponto onde se pretende calcular o deslocamento (ou um momento unitário no ponto onde se pretende calcular a rotação).

A aplicação do **Teorema da Carga Unitária** envolve a determinação das expressões doa diagramas de momento fletor na estrutura para dois cenários de carregamento: (i) o carregamento real; (ii) o carregamento virtual. Considerando apenas o efeito do momento fletor na deformação da estrutura (desprezando-se, desta forma, o contributo do esforço transverso), resulta a seguinte expressão:

\begin{equation*}
1 \cdot \delta_{D} = \sum_i \int_0^{L_i} \frac{M(x)M_u(x)}{EI} \mathrm{d}x
\end{equation*}

### Resolução

- Cálculo das reações
\begin{equation*}
\sum F_x = 0
\quad\wedge\quad
\sum F_y = 0
\quad\wedge\quad
\sum M_A = 0
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

QB: float = 40.e3 # unit N/m
LAB: float = 2. # unit m
LBC: float = 2. # unit m
LCD: float = 3. # unit m

rax, ray, rdy = sy.symbols('rax ray rdy')

print('::: sistema de equações :::')
sumFX = rax
print(f'{sumFX = } [N]')
sumFY = ray + rdy - QB*LAB/2 - QB*LBC/2
print(f'{sumFY = } [N]')
sumMA= rdy*(LAB+LBC) - QB*LAB/2*(2/3*LAB) - QB*LBC/2*(LAB+1/3*LBC)
print(f'{sumMA = } [N.m]')
sol = solve([sumFX,sumFY,sumMA],[rax, ray, rdy])
```

```python
print('::: resolução :::')
print(sol)
RAx = sol[rax]; print(f'{RAx = :.1f} [kN]')
RAy = sol[ray]; print(f'{RAy = :.1f} [kN]')
RDy = sol[rdy]; print(f'{RDy = :.1f} [kN]')
```

#### Diagrama de esforços devido ao carregamento real

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au10/P8/MSII_Au10_P8_i.png"
style="width: 40%;"/>
  
- **Tramo AB:**

Expressão do momento fletor:

\begin{equation*}
w_1 (x_1) = \frac{Q_B}{L_{AB}}x_1 = 20 x_1
\quad\wedge\quad x_1 \in [A, B[
\end{equation*}

\begin{equation*}
\sum M  = 0
\quad\therefore\quad
-R_{Ay}x_1 + \left(\int_0^{x_1} \frac{Q_B}{L_{AB}}\overline{x}_1 \mathrm{d}\overline{x}_1 \right)
\left(\frac{1}{3}x_1\right) + M_{AB}(x_1) = 0
\end{equation*}

\begin{equation*}
\begin{aligned}
M_{AB}(x_1) &= - \frac{Q_B}{L_{AB}} \left[\frac{x_1^2}{2}\right]_0^{x_1} 
\left(\frac{1}{3}x_1\right) + R_{Ay}x_1 \\
&= -\frac{Q_B}{6L_{AB}} x_1^3 + R_{Ay}x_1 \\
&= -\frac{10}{3} x_1^3 + 40x_1
\end{aligned}
\end{equation*}


```python
x = sy.symbols('x')
MAB = - sy.integrate(QB/LAB*x,(x,0,x))*x/3 + RAy*x
print(f'{MAB = } [N.m]')
```

- **Tramo CB:**

Expressão do momento fletor:

\begin{equation*}
M_{CB}(x) = -\frac{Q_C}{6L_{CB}} x_2³ + R_{Ay}x_2
= -\frac{10}{3} x_2³ + 40x_2    
\quad\wedge\quad x_2 \in [C, B[
\end{equation*}


```python
MCB = - sy.integrate(QB/LAB*x,(x,0,x))*x/3 + RAy*x
print(f'{MCB = } [N.m]')
```

### Aplicação do Teorema da Carga Unitária: Diagrama de esforços devido ao carregamento unitário

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au10/P8/MSII_Au10_P8_ii.png"
style="width: 40%;"/>

Aplica-se à estrutura uma força unitária no ponto e direção onde o deslocamento se pretende determinar, neste caso, no nó D e na direção $x$. Para este novo carregamento unitário, determina-se as reações e as expressões do momento fletor (pela aplicação do método das secções).


```python
raxu, rayu, rdyu = sy.symbols('raxu rayu rdyu')

Pu = 1

print('::: sistema de equações :::')
sumFX = raxu + Pu
print(f'{sumFX = } [N]')
sumFY = rayu + rdyu
print(f'{sumFY = } [N]')
sumMA= -rdyu*(LAB+LBC) + Pu*LCD
print(f'{sumMA = } [N.m]')
sol = solve([sumFX,sumFY,sumMA],[raxu, rayu, rdyu])
```

```python
print(sol)
print('::: resolução :::')
RAxu = sol[raxu]; print(f'{RAxu = :.1f} [kN]')
RAyu = sol[rayu]; print(f'{RAyu = :.2f} [kN]')
RDyu = sol[rdyu]; print(f'{RDyu = :.2f} [kN]')
```

- **Tramo AB (carga unitária) :**

Expressão do momento fletor:

\begin{equation*}
x_1 \in [A, B[ ~: \quad
\sum M  = 0
\Leftrightarrow
-R_{Ayu}x_1 + M_{ABu}(x_1) = 0
\end{equation*}

resultando,

\begin{equation*}
M_{ABu}(x_1) = R_{Ayu}x_1 = 0.75x_1
\end{equation*}


```python
MABu = -RAyu*x
print(f'{MABu = } [N.m]')
```

- **Tramo CB (carga unitária):**

Expressão do momento fletor:

\begin{equation*}
x_2 \in [C, B[ ~: \quad
\sum M  = 0
\Leftrightarrow
1 \cdot L_{CD} - R_{Dyu}x_2 - M_{CBu}(x) = 0
\end{equation*}

resultando,

\begin{equation*}
M_{CBu}(x) = L_{CD} - R_{Dyu}x_2 = 3 - 0.75x_2
\end{equation*}

```python
MCBu = LCD+RAyu*x
print(f'{MCBu = } [N.m]')
```

### Cálculo do deslocamento horizontal no ponto D

- Teorema da Carga Unitária

\begin{equation*}
1 \cdot \delta_{D} = \sum_i \left.\int_0^{L_i} \frac{M(x)M_u(x)}{EI}\right|_i\mathrm{d}x ,
\quad i = 1, 2, 3 
\end{equation*}

- **Troço AB:**

\begin{equation*}
x_1 \in [A,B[ : \quad
\delta_{D_1} = \int_{x_A}^{x_B} \frac{M_{AB}(x_1)M_{ABu}(x_1)}{EI} \mathrm{d}x =
\frac{1}{EI}\int_{0}^{2}\left(-\frac{10}{3} x_1³ + 40x_1\right)(0.75x_1) \mathrm{d}x
\end{equation*}

- **Troço CB:**

\begin{equation*}
x_2 \in [C,B[ : \quad
\delta_{D_2} = \int_{x_C}^{x_B} \frac{M_{CB}(x_2)M_{CBu}(x_2)}{EI}\mathrm{d}x =
\frac{1}{EI}\int_{0}^{2}\left(-\frac{10}{3} x_2³ + 40x_2\right)(3 - 0.75x_2) \mathrm{d}x
\end{equation*}

- **Troço DC:**

\begin{equation*}
x_3 \in [D,C[ : \quad
M_{DC}(x_3) = 0 \quad\therefore\quad \delta_{D_3} = 0
\end{equation*}

Pelo que o deslocamento será dado por,

\begin{equation*}
\delta_{D} = \delta_{D_1} + \delta_{D_2}  = 2.116\times10^{-1}~\textrm{m}
\end{equation*}

```python
E: float = 210.e9 # unit : Pa
I: float = 4.5e6*1e-12 # unit : m⁴
EI = E*I
print(f'EI = {EI} [N.m²]')
deltaD1 = sy.integrate(MAB*MABu,(x,0,LAB))/EI
print(f'deltaD1 = {sy.integrate(MAB*MABu,(x,0,LAB)):.1f}/EI = {deltaD1:.3e} [m]')
deltaD2 = sy.integrate(MCB*MCBu,(x,0,LBC))/EI
print(f'deltaD1 = {sy.integrate(MCB*MCBu,(x,0,LBC)):.1f}/EI = {deltaD2:.3e} [m]')
deltaD = deltaD1 + deltaD2
print(f'deltaD = {deltaD:.3e} [m] = {deltaD*1e-3:.3e} [mm]')
```

# Anexo I:

## Teorema da Carga Unitária

O Teorema da Carga Unitária é um método para calcular deslocamentos (ou rotações) em estruturas usando o Princípio dos Trabalhos Virtuais. A ideia fundamental é:

**Para calcular o deslocamento num ponto e direção específicos:**
1. Aplica-se uma carga unitária fictícia (força = 1 ou momento = 1) no ponto e direção onde se pretende conhecer o deslocamento
2. Calcula-se o diagrama de momentos fletores devido a esta carga unitária: **Mu(x)**
3. Calcula-se o diagrama de momentos fletores devido ao carregamento real: **M(x)**
4. Integra-se o produto dos dois diagramas ao longo da estrutura

Como mostra a expressão na tua imagem:

$$1 \cdot \delta_D = \sum_i \int_0^{L_i} \frac{M(x)M_u(x)}{EI} \, dx$$

Onde:
- **δD** = deslocamento pretendido
- **M(x)** = momento fletor devido ao carregamento real
- **Mu(x)** = momento fletor devido à carga unitária
- **EI** = rigidez à flexão

## Teorema de Castigliano

O **Segundo Teorema de Castigliano** (o mais usado) estabelece que:

> *A derivada parcial da energia de deformação em relação a uma força (ou momento) aplicada é igual ao deslocamento (ou rotação) no ponto de aplicação dessa força, na sua direção.*

Matematicamente:

$$\delta_i = \frac{\partial U}{\partial P_i}$$

Onde:
- **U** = energia de deformação total da estrutura
- **Pi** = força aplicada no ponto i
- **δi** = deslocamento no ponto i na direção de Pi

Para flexão, a energia de deformação é:

$$U = \sum_i \int_0^{L_i} \frac{M(x)^2}{2EI} \, dx$$

## Relação entre os Teoremas

**Os dois teoremas são matematicamente equivalentes!** A demonstração vem de:

1. Pelo Teorema de Castigliano:
$$\delta = \frac{\partial U}{\partial P} = \frac{\partial}{\partial P} \left[\int \frac{M(x)^2}{2EI} dx\right]$$

2. Aplicando a derivada (e lembrando que M depende de P):
$$\delta = \int \frac{M(x)}{EI} \cdot \frac{\partial M(x)}{\partial P} \, dx$$

3. Mas $\frac{\partial M(x)}{\partial P}$ é exatamente o momento fletor causado por uma carga unitária aplicada em P, ou seja, **Mu(x)**!

4. Logo:
$$\delta = \int \frac{M(x) \cdot M_u(x)}{EI} \, dx$$

Que é precisamente a expressão do Teorema da Carga Unitária.

## Diferenças Práticas

| **Aspecto** | **Castigliano** | **Carga Unitária** |
|-------------|-------------------|-----------------|
| **Quando usar** | Quando já existe uma força no ponto de interesse | Quando não há força no ponto (adiciona-se uma fictícia) |
| **Procedimento** | Derivação da energia | Dois estados: real + unitário |
| **Vantagem** | Conceitualmente mais simples | Pode ser mais direto para estruturas complexas |
