### Problema 9

Considerando a Figura, mostre que se uma viga de secção transversal retangular estiver submetida a um binário aplicado num plano que contenha uma diagonal da secção transversal, o eixo neutro estará dirigido segundo a direção da outra diagonal.

### 

Considere a seguinte expressão:

\begin{equation*}
\tan(\phi) = \frac{I_z}{I_y}\tan(\theta)
\end{equation*}

em que, 

- $\phi$ é o ângulo entre o eixo $z$ e o eixo neutro, 
- $\theta$ é o ângulo entre o eixo $z$ e o momento fletor aplicado $M$.

Desenvolvendo ,


\begin{equation*}
\tan(\phi) = \frac{h³b/12}{bh³/12} = \frac{h}{b} 
\end{equation*}


como $\tan(\phi) = h/b$, então o ângulo entre $\overline{AD}$ e $\overline{DE}$ é $\theta$.
