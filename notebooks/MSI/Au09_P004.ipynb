```python
from numpy import pi, sqrt, arctan, rad2deg, sin

# unidades: mm. N, MPa

# propriedades elásticas:
E = 200.e3 # unidade: MPa

# geometria (dimensões):
phi = 2. # unidade: mm
A = pi*phi**2/4
print(f'A = {A:.4f} mm^2')
LAC = 300. # unidade: mm
LCB = LAC
LAD = 300.

P = 750. # unidade: N
```

#### Equilíbrio das barras (estrutura)

De notar que existe simetria no problema: (i) geométrica; (ii) de carregamento; (iii) de propriedades mecânicas, em relação a um eixo vertical que contem o ponto $C$. Pela análise do equilíbrio da barra rígida $ACB$ conclui-se:

\begin{equation*}
\begin{aligned}
(+)\uparrow~&:~\sum F_y=0
\quad\Leftrightarrow\quad
2F_1 + 2F_2\sin\theta - P =0 \\
  \end{aligned}
\end{equation*}

O valor do ângulo $\theta$ é calculado atendendo à configuração inicial da estrutura:

\begin{equation*}
\theta = \arctan\left(\frac{L_{AD}}{L_{AC}}\right) = 45^\circ~;~
\sin(45^\circ) = \frac{\sqrt{2}}{2}
\end{equation*}

pelo que se conclui:

\begin{equation*}
2F_1 + \sqrt{2}F_2 - P =0
~\Leftrightarrow~
F_1 = \frac{P-\sqrt{2}F_2}{2}
\end{equation*}

Este problema é estaticamente indeterminado. Para a sua resolução, é necessário introduzir informação adicional pela análise física do problema. Esta é definida pela contabilidade de deslocamentos gerada pela hipótese da barra $ACB$ ser rígida (deformação desprezável). Da geometria e simetria do problema, resulta a seguinte equação de compatibilidade de deslocamentos:

\begin{equation*}
\delta_A = \delta_B = \delta_C
\end{equation*}

Como a variação de comprimento é de origem mecânica, o seu valor é proporcional à força axial que se instalam nos cabos. Para os cabos verticais têm-se:

\begin{equation*}
\delta_A = \delta_B = \frac{F_1 L_{AD}}{AE}
\end{equation*}

Para os cabos inclinados a deformação axial tem que levar em consideração a orientação inclinada dos mesmos ($\theta=45^\circ$):

\begin{equation*}
\delta_{CE} = \delta_C \sin\theta
~\Leftrightarrow~
\delta_C = \frac{\delta_{CE}}{\sin(45^\circ)} = \frac{F_2 L_{CE}}{AE}
\left(\frac{2}{\sqrt{2}}\right) = \frac{\sqrt{2} F_2 L_{CE}}{AE}
\end{equation*}

Atendendo à relação geométrica:

\begin{equation*}
L_{AD} = L_{CE}\sin(45^\circ)
= L_{CE} \left(\frac{\sqrt{2}}{2}\right)
\end{equation*}

a compatibilidade de deslocamentos resulta finalmente:

\begin{equation*}
\delta_A = \delta_C
~\Leftrightarrow~
\frac{F_1 L_{AD}}{AE} = \frac{\sqrt{2}F_2 L_{CE}}{AE}
~\Leftrightarrow~
F_1 = 2 F_2
\end{equation*}

#### sistema de equações

Pela análise desenvolvida é possível obter o seguinte sistema de equações:

\begin{equation*}
\begin{cases}
F_1 ~&= \displaystyle\frac{P-\sqrt{2}F_2}{2} \\
F_2 ~&= F_1/2
\end{cases}
~\Leftrightarrow~
\begin{cases}
F_1 ~&= \displaystyle\frac{2P}{4+\sqrt{2}} \\
F_2 ~&=\displaystyle\frac{P}{4+\sqrt{2}}
\end{cases}
\end{equation*}

```python
F1 = 2*P/(4+sqrt(2))
print(f'F1 = {F1:.2f} N')

F2 = F1/2
print(f'F2 = {F2:.2f} N')
```

```python
sigA = F1/A
print(f'sigA = {sigA:.2f} MPa')

sigC = F2/A
print(f'sigC = {sigC:.2f} MPa')
```
