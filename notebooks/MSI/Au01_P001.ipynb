## Problema 1 <a name="PI11"></a>


Considere o estado de tensão num ponto $p$, representado pelo seguinte tensor, no
referencial cartesiano $S(0,x,y,z)$.

\begin{equation*}
\underline{\underline{\sigma}}^T =
\begin{bmatrix}
1000 & 200 & 0 \\
200 & -6000 & -400 \\
0 & -400 & 0
\end{bmatrix}~\textrm{(kPa)}
\end{equation*}

a) Calcule o vetor tensão aplicado num plano perpendicular ao versor $\vec{n} = (0,11; 0,35; 0,93)$ e que passa por $p$.

b) Calcule as componentes de tensão normal e tangencial a esse plano.

#### Nomenclatura

\begin{equation*}
\begin{aligned}
\vec{t}(p,\vec{n}) &\quad \textrm{vetores tensão que atua no ponto P e numa faceta normal a $\vec{n}$} \\
\vec{n} & \quad \textrm{versor exterior à superfície que comtem o ponto material P em análise} \\
\tilde{\sigma} & \quad \textrm{tensor das tensões} \\
\end{aligned}
\end{equation*}

## Resolução

#### a) Calcule o vetor tensão aplicado num plano perpendicular ao versor $\vec{n} = (0,11; 0,35; 0,93)$ e que passa por $p$.

O princípio das tensões de Cauchy estabelece que as interações materiais que ocorrem em superfícies internas de um corpo podem ser descritas como uma distribuição de forças de tração, da mesma forma como é descrito o efeito de forças externas que atuam nas superfícies físicas do corpo.

A partir da representação matricial do tensor das tensões num determinado ponto $p$, $(\tilde{\sigma} \equiv \underline{\underline{\sigma}})$, é possível determinar o vetor tensão, $\vec{t}{(p,\vec{n})}$, que atua nesse ponto sobre uma faceta (plano) cuja orientação no espaço é definida pelo versor exterior à superfície, $\vec{n}$. Essa relação é conhecida pela fórmula da tensão de Cauchy, sendo expressa por:

\begin{equation*}
\begin{aligned}
\vec{t}{(p,\vec{n})} &= \tilde\sigma^T(p) \cdot \vec{n}
\quad|\quad \underline{t} (p,\vec{n}) = \underline{\underline{\sigma}}^T(p) \cdot \underline{n}
&\quad\textrm{(notação direta)} \\
t_i &= \sigma_{ji} \cdot n_j &\quad\textrm{(notação indicial)}
\end{aligned}
\end{equation*}

Num sistema de coordenadas $S(0,\vec{e}_1,\vec{e}_2,\vec{e}_3)$ e para efeitos de cálculo pode usar-se a seguinte expressão matricial:

\begin{equation*}
\begin{aligned}
\begin{bmatrix}
t_1 \\ t_2 \\ t_3
\end{bmatrix} &=
\begin{bmatrix}
\sigma_{11} & \sigma_{21} & \sigma_{31} \\
\sigma_{12} & \sigma_{22} & \sigma_{32} \\
\sigma_{13} & \sigma_{23} & \sigma_{23}
\end{bmatrix}
\begin{bmatrix}
n_1 \\ n_2 \\ n_3
\end{bmatrix}
\end{aligned}
\end{equation*}

em que, cada coluna da matriz $\underline{\underline{\sigma}}^T(p)$ é obtida pelas componentes cartesianas dos vetores tensão, $\vec{t}(\vec{e}_j)$, que atuam nas facetas normais aos eixos coordenados do sistema ortonormado e direto:

\begin{equation*}
\begin{cases}
\vec{t}(\vec{e}_1) = \sigma_{11} \vec{e}_1 + \sigma_{12} \vec{e}_2 + \sigma_{13} \vec{e}_3\\
\vec{t}(\vec{e}_2) = \sigma_{21} \vec{e}_1 + \sigma_{22} \vec{e}_2 + \sigma_{23} \vec{e}_3\\
\vec{t}(\vec{e}_3) = \sigma_{31} \vec{e}_1 + \sigma_{32} \vec{e}_2 + \sigma_{33} \vec{e}_3\\
\end{cases}
\end{equation*}

Em alternativa, expressando o vetor tensão como um vetor coluna, o cálculo das componentes do vetor tensão num ponto e numa determinada faceta pode ser obtido por:

\begin{equation*}
\vec{t}^{~T} \equiv
\begin{bmatrix}
t_1 & t_2 & t_3
\end{bmatrix} =
\begin{bmatrix}
n_1 & n_2 & n_3
\end{bmatrix}
\begin{bmatrix}
\sigma_{11} & \sigma_{12} & \sigma_{13} \\
\sigma_{21}(=\sigma_{12}) & \sigma_{22} & \sigma_{23} \\
\sigma_{31}(=\sigma_{13}) & \sigma_{32}(=\sigma_{23}) & \sigma_{23}
\end{bmatrix}
\end{equation*}

```python

%matplotlib inline
import numpy as np

def versor(n1,n2,n3):
    return np.array([n1, n2, n3])

def tensortensoes(s11,s12,s13,s21,s22,s23,s31,s32,s33):
    return np.array([[s11,s12,s13], [s21,s22,s23], [s31,s32,s33]])

versor_n = versor(0.11, 0.35, 0.93)
Sigma_n = tensortensoes(1000, 200, 0, 200, -6000, -400, 0, -400, 0)
vetor_tensao_n = np.dot(versor_n,Sigma_n)

print("O vetor tensão no ponto p num plano perpendicular ao versor (0,11; 0,35; 0,93) é :")
print("t^n = ",vetor_tensao_n, "kPa")
```

#### b) Calcule as componentes de tensão normal e tangencial a esse plano.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au01/Au01_P001_i1.jpg" alt="DEMI" width="280" height="220">

O vetor de tensão pode ser obtido pelo sumários das seguintes componentes, num referencial local associado à orientação da faceta no ponto $p$:

\begin{equation*}
\vec{t}(p,\vec{n}) = \vec{\sigma_n} + \vec{\tau} = \sigma_n~\vec{n} + \tau~\vec{s}
\quad \textrm{com} \quad \vec{\sigma_n} \cdot \vec{\tau} = 0
\end{equation*}

em que $\sigma_n$ é a tensão normal e $\tau$ é a tensão tangencial, no ponto $p$ e para a direção definida por $\vec{n}$.

A componente de tensão normal do vetor de tensão no ponto $p$ e segundo a normal externa à faceta, $\sigma_n$, pode ser cálculda pelo produto interno assim definido:

\begin{equation*}
\sigma_n(p,\vec{n}) = \vec{t} \cdot \vec{n} = t_i \cdot n_i = \{t\}^{~T} \{n\} =
\begin{bmatrix}
t_x & t_y & t_z
\end{bmatrix}
\begin{bmatrix}
n_x \\ n_y \\ n_z
\end{bmatrix} = t_x n_x + t_y n_y + t_z n_z
\end{equation*}

A tensão normal diz-se de tração se for positiva ou de compressão no caso negativo.

```python
tensao_normal = np.dot(vetor_tensao_n,versor_n)
print("A componente de tensão normal é: ",'{:.2f}'.format(tensao_normal), "kPa")
```

Por sua vez, a componente de tensão tangencial do vetor de tensão, $\tau$, pode ser simplesmente cálculado atendendo ao teorema de Pitágoras:

\begin{equation*}
||\vec{t}||^2 = \vec{t}\cdot\vec{t} = \sigma_n^2 + \tau^2
\end{equation*}

desenvolvendo algebricamente,

\begin{equation*}
\tau(p,\vec{n}) = ||\vec{t}^{(\vec{n})} - \vec{\sigma}_n|| =
\sqrt{\vec{t}\cdot\vec{t} - \sigma_n^2} = \sqrt{t_x^2 + t_y^2 + t_z^2 - \sigma_n^2} \quad \textrm{;} \quad
\vec{\tau}(P,\vec{n}) \equiv \tau\vec{s} = \vec{t} - \vec{\sigma}_n
\end{equation*}

```python
vt_n2 = np.dot(vetor_tensao_n,vetor_tensao_n)
tensao_tangencial = np.sqrt(vt_n2 - tensao_normal**2)
print(f"A componente da tensão tangencial é: {tensao_tangencial:.2f} kPa")
```
