## Problema 8 <a name="PI28"></a>

Considere o estado de tensão num ponto $P$ de um sólido, relativo a um sistema de coordenadas $S_0(0,x_0,y_0,z_0)$, definido pela seguinte matriz das tensões:

\begin{equation*}
\sigma_{ji} = [\sigma]^\top = \begin{bmatrix}
-8 & 6 & -2 \\
6 & 4 & 2 \\
-2 & 2 & -5
\end{bmatrix}~\textrm{MPa}
\end{equation*}

Determine o estado de tensão num elemento que está orientado em relação so sitema de coordenadas $S_0$ por uma rotação de $\theta=45^\circ$ segundo o eixo dos $zz$, seguido de uma rotação de $\phi=30^\circ$ segundo o eixo dos $xx$.

## Resolução

A matriz de transformação entre $S_0(0,x_0,y_0,z_0)$ e $S_1(0,x_1,y_1,z_1)$ definida por uma rotação de $\theta=45^\circ$ segundo o eixo dos $zz$ é dada por:

\begin{equation*}
\begin{aligned}
\{n\}_1 &= [T]_{0/1}\{n\}_0 \\
\begin{bmatrix}
x_1 \\ y_1 \\ z_1
\end{bmatrix} &= %
\begin{bmatrix}%
\cos(x_0,x_1) & \cos(y_0,x_1) & \cos(z_0,x_1) \\
\cos(x_0,y_1)  & \cos(y_0,y_1)  & \cos(z_0,y_1) \\
\cos(x_0,z_1)  & \cos(y_0,z_1)  & \cos(z_0,z_1) \\
\end{bmatrix}%
\begin{bmatrix}%
x_0 \\ y_0 \\ z_0
\end{bmatrix}
\end{aligned}
\end{equation*}

desenvolvendo algebricamente,

\begin{equation*}
\begin{aligned}
\begin{bmatrix}
x_1 \\ y_1 \\ z_1
\end{bmatrix} &= %
\begin{bmatrix}%
\cos(\theta) & \cos(\pi/2-\theta) & \cos(\pi/2) \\
\cos(\pi/2+\theta)  & \cos(\theta)  & \cos(\pi/2) \\
\cos(\pi/2)  & \cos(\pi/2)  & \cos(0) \\
\end{bmatrix}%
\begin{bmatrix}%
x_0 \\ y_0 \\ z_0
\end{bmatrix}
\end{aligned}
\end{equation*}

resultando,

\begin{equation*}
\begin{aligned}
\begin{bmatrix}
x_1 \\ y_1 \\ z_1
\end{bmatrix} &= %
\begin{bmatrix}%
\cos\theta & \sin\theta & 0 \\
-\sin\theta  & \cos\theta  & 0 \\
0 & 0  & 1 \\
\end{bmatrix}%
\begin{bmatrix}%
x_0 \\ y_0 \\ z_0
\end{bmatrix}
\end{aligned}
\end{equation*}

De igual forma a matriz de transformação entre $S_1(0,x_1,y_1,z_1)$ e $S_2(0,x_2,y_2,z_2)$ definida por uma rotação de $\phi=30^\circ$ segundo o eixo dos $xx$ é dada por:

\begin{equation*}
\begin{aligned}
\{n\}_2 &= [T]_{1/2}\{n\}_1 \\
\begin{bmatrix}
x_2 \\ y_2 \\ z_2
\end{bmatrix} &= %
\begin{bmatrix}%
1 & 0 & 0 \\
0 & \cos\phi & \sin\phi \\
0 & -\sin\phi & \cos\phi \\
\end{bmatrix}%
\begin{bmatrix}%
x_1 \\ y_1 \\ z_1
\end{bmatrix}
\end{aligned}
\end{equation*}

Pelo que a transformação entre $S_0(0,x_0,y_0,z_0)$ e $S_2(0,x_2,y_2,z_2)$ vem:

\begin{equation*}
\begin{aligned}
\{n\}_2 &= [T]_{0/2}\{n\}_0 =
[T]_{1/2}\underbrace{[T]_{0/1}\{n\}_0}_{\{n\}_1} \\[2ex]
\begin{bmatrix}%
x_2 \\ y_2 \\ z_2
\end{bmatrix}%
&=\begin{bmatrix}%
1 & 0 & 0 \\
0 & \cos\phi & \sin\phi \\
0 & -\sin\phi & \cos\phi \\
\end{bmatrix}%
\begin{bmatrix}%
\cos\theta & \sin\theta & 0 \\
-\sin\theta  & \cos\theta  & 0 \\
0 & 0  & 1 \\
\end{bmatrix}
\begin{bmatrix}%
x_0 \\ y_0 \\ z_0
\end{bmatrix}
\end{aligned}
\end{equation*}

cuja multiplicação de matriz resulta em,

\begin{equation*}
\begin{aligned}
\begin{bmatrix}%
x_2 \\ y_2 \\ z_2
\end{bmatrix} =
\begin{bmatrix}%
\cos\theta & \sin\theta & 0 \\
-\sin\theta\cos\phi & \cos\theta\cos\phi  & \sin\phi \\
\sin\theta\sin\phi & -\cos\theta\sin\phi  & \cos\phi \\
\end{bmatrix}%
\begin{bmatrix}%
x_0 \\ y_0 \\ z_0
\end{bmatrix}
\end{aligned}
\end{equation*}

A matriz das tensões do referêncial $S_2(0,x_2,y_2,z_2)$ pode obter-se à custa das componentes do tensor das tensões no sitema de coordeandas $S_0(0,x_0,y_0,z_0)$ através da seguinte lei de transformação:

\begin{equation*}
[\sigma]_2 = [T]_{0/2} \,[\sigma]_0 \, [T]_{0/2}^T
\end{equation*}

```python
import numpy as np
np.set_printoptions(precision=3)

def matrizT(l1,m1,n1,l2,m2,n2,l3,m3,n3):
    return np.array([[l1,l2,l3],[m1,m2,m3],[n1,n2,n3]])

def matrizTensoes(sxx,sxy,sxz,syy,syz,szz):
    syx = sxy; szx = sxz; szy = syz
    return np.array([[sxx,sxy,sxz],[syx,syy,syz],[szx,szy,szz]])

exx = 0.004 
eyy = -0.004 
exy = 0.005
ezz = 0.018



ang01 = np.deg2rad(45)
T01 = matrizT(np.cos(ang01),-np.sin(ang01),0,np.cos(ang01),np.sin(ang01),0,0,0,1)
print('\nT01 = ', T01)

ang12 = np.deg2rad(30)
T12 = matrizT(1,0,0,0,np.cos(ang12),-np.sin(ang12),0,np.sin(ang12),np.cos(ang12))
print('\nT12 = ', T12)

T02 = np.dot(T12,T01)
print('\nT02 = ', T02)

mt0 = matrizTensoes(-8,6,-2,4,2,-5)
print('\nmt0 = ', mt0)

mt2 = np.dot(T02,np.dot(mt0,np.transpose(T02)))
print('\nmt2 = ', mt2)
```

Esta operação de transformação pode ser verificada através dos invariantes do tensor das tensões:

\begin{equation*}
\begin{aligned}
I_1 &= \mathrm{Tr}(\,\sigma_{{ij/0}}\,) = \mathrm{Tr}(\,\sigma_{ij/2}\,)\\
I_2 &= \frac{1}{2}\left[ \mathrm{Tr}(\,\sigma_{ij/0}\,)^2 - \mathrm{Tr}(\,\sigma_{ij/0}^2)\, \right]
= \frac{1}{2}\left[ \mathrm{Tr}(\,\sigma_{ij/2}\,)^2 - \mathrm{Tr}(\,\sigma_{ij/2}^2)\, \right]\\
I_3 &= \mathrm{det}(\,\sigma_{ij/0}\,) = \mathrm{det}(\,\sigma_{ij/2}\,) \\
\end{aligned}
\end{equation*}

```python
I1_S0 = np.trace(mt0)
print('1º Invariante, I1 - S0 :','{:.1f}'.format(I1_S0))

I1_S2 = np.trace(mt2)
print('1º Invariante, I1 - S2 :','{:.1f}'.format(I1_S2))

I2_S0 = 0.5*(np.trace(mt0)**2-np.trace(np.dot(mt0,mt0)))
print('\n2º Invariante, I2 - S0:','{:.1f}'.format(I2_S0))

I2_S2 = 0.5*(np.trace(mt2)**2-np.trace(np.dot(mt2,mt2)))
print('2º Invariante, I2 - S2:','{:.1f}'.format(I2_S2))

I3_S0 = np.linalg.det(mt0)
print('\n3º Invariante, I3 - S0:','{:.1f}'.format(I3_S0))

I3_S2 = np.linalg.det(mt2)
print('3º Invariante, I3 - S2:','{:.1f}'.format(I3_S2))
```

```python
I1 = I1_S0; I2 = I2_S0; I3 = I3_S0

print('\n Resolucao Numerica - Numpy : ............................')
valpropriosN = np.roots([1,-I1,I2,-I3])
print('\n Numpy : roots = ',valpropriosN)

srange = np.linspace(np.min(valpropriosN)*1.2,np.max(valpropriosN)*1.2,50)
def polycaract(s):
    return s**3 - I1*s**2 + I2*s - I3

AAA = polycaract(srange)
import matplotlib.pyplot as plt
font = {'size'   : 15}
plt.rc('font', **font)
plt.rcParams['axes.facecolor']='black'
plt.figure(figsize=(9,6))
plt.plot(srange,polycaract(srange),linewidth=3)
for i in valpropriosN:
    plt.plot(i,polycaract(i),'o',markersize=20)
plt.ylabel('Poly caract. ')
plt.xlabel('$\sigma$')
plt.grid('on')
plt.show()
```
