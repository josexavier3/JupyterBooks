## Problema 4 <a name="PI14"></a>

Calcule a tensão normal e de corte na faceta que faz um ângulo de 30$^\circ$ com a horizontal, como indicado na figura. A peça tem uma secção com 2x2 centímetros (cm) e está sujeita a uma carga de compressão de 13 kN.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au01/Au01_P004_i1.jpg" width="300" height="475">


## Resolução

O tensor das tensões representa neste caso um estado de tensão uniaxial e uniforme representado por:

\begin{equation*}
\underline{\underline{\sigma}}^\top = \begin{bmatrix}
0 & 0 & 0 \\
0 & \sigma_{yy} & 0 \\
0 & 0 & 0
\end{bmatrix}
\end{equation*}

em que $\sigma_{yy} = P/A$, sendo $P$ a força de compresão aplicada e $A$ a área de secção transversal inicial do provete.

Neste exemplo assume-se que a direção da aplicação da carga compressiva ocorre segundo o eixo $y$. A faceta do elemento roda de um ângulo $\theta$ no plano ($y,z$) em relação ao eixo comum $x$. O vector normal à faceta é portanto dado por:

\begin{equation*}
\vec{n} \equiv \underline{n} \equiv n_i = \begin{bmatrix}
n_1 & n_2 & n_3 \\
\end{bmatrix} =
\begin{bmatrix}
0 & \cos\theta & \sin\theta
\end{bmatrix}
\end{equation*}

Desta forma o vector tensão que atua num ponto da faceta é dado por:

\begin{equation*}
\begin{bmatrix}
t_1 \\ t_2 \\ t_3
\end{bmatrix} =
\begin{bmatrix}
\sigma_{11} & \sigma_{21} & \sigma_{31} \\
\sigma_{12} & \sigma_{22} & \sigma_{32} \\
\sigma_{13} & \sigma_{23} & \sigma_{33}
\end{bmatrix}
\begin{bmatrix}
n_1 \\ n_2 \\ n_3
\end{bmatrix}
\end{equation*}

ou, alternativamente:

\begin{equation*}
\begin{aligned}
\begin{bmatrix}
t_1 & t_2 & t_3
\end{bmatrix} 
=
\begin{bmatrix}
n_1 & n_2 & n_3
\end{bmatrix}
\begin{bmatrix}
\sigma_{11} & \sigma_{12} & \sigma_{13} \\
\sigma_{21}(=\sigma_{12}) & \sigma_{22} & \sigma_{23} \\
\sigma_{31}(=\sigma_{13}) & \sigma_{32}(=\sigma_{23}) & \sigma_{23}
\end{bmatrix}
\end{aligned}
\end{equation*}

\begin{equation*}
\begin{aligned}
\begin{bmatrix}
t_1 & t_2 & t_3
\end{bmatrix}
=
\begin{bmatrix}
0 & \cos\theta & \sin\theta
\end{bmatrix}
\begin{bmatrix}
0 & 0 & 0 \\
0 & \sigma_{yy} & 0 \\
0 & 0 & 0
\end{bmatrix} = 
\begin{bmatrix}
0 & \sigma_{yy}\cos\theta & 0
\end{bmatrix}
\end{aligned}
\end{equation*}

Em particular, para $theta = 30^\circ$ tem-se:

\begin{equation*}
\vec{n}(\theta=30^\circ) =
\begin{bmatrix}
0 & \displaystyle\frac{\sqrt{3}}{2} & \displaystyle\frac{1}{2}
\end{bmatrix} \quad \textrm{e} \quad \vec{t}^{(\vec{n})} =
\begin{bmatrix}
0 & \displaystyle\frac{\sqrt{3}P}{2A} & 0
\end{bmatrix}
\end{equation*}

E, para $\theta = 45^\circ$ vem:

\begin{equation*}
\vec{n}(\theta=45^\circ) =
\begin{bmatrix}
0 & \displaystyle\frac{\sqrt{2}}{2} & \displaystyle\frac{\sqrt{2}}{2}
\end{bmatrix} \quad \textrm{e} \quad \vec{t}^{(\vec{n})} =
\begin{bmatrix}
0 & \displaystyle\frac{\sqrt{2}P}{2A} & 0
\end{bmatrix}
\end{equation*}

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

P = -13e3 # unit: N
A0 = 20*20 # unit: mm
theta = 30 # unit: degree
theta2 = 45 # unit: degree
Syy = P/A0
angrange = np.arange(0,91,5) # unit: degree

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#plt.figure(dpi=300)
ax.plot(angrange,Syy*np.cos(np.deg2rad(angrange)), color='k', linestyle='-', linewidth=3.0, marker="o",
         markersize=10, label = r'$\theta$')
ax.set_xlabel(r'Angulo (deg)')
ax.set_ylabel(r'$\sigma_{yy}\cos\theta$', color='k')
ax.set_xlim([0, 100]); ax.set_ylim([Syy, 10])
ax.grid()

print(' Tensao de compressao aplicada: Syy = P/A = ', '{:.2f}'.format(Syy), 'MPa')
print(' Tensao de compressao na direcao n2 (30Deg) : Syy = (P/A)cos(ang) = ',
      '{:.2f}'.format(Syy*np.cos(np.deg2rad(theta))), 'MPa')
print(' Tensao de compressao na direcao n2 (45Deg) : Syy = (P/A)cos(ang) = ',
      '{:.2f}'.format(Syy*np.cos(np.deg2rad(theta2))), 'MPa')
```

À componente do vetor das tensões, $\vec{t}^{(\hat{n})}\equiv \vec{\sigma}^{(\vec{n})}$. segundo a normal à faceta designa-se por tensão normal, $\sigma_n$, e pode ser obtida fazendo o produto interno de $\vec{t}^{(\vec{n})}$ por $\vec{n}$:

\begin{equation*}
\sigma_n = \vec{t}^{(\vec{n})} \cdot \vec{n} =
\begin{bmatrix}
0 & \sigma_{yy}\cos\theta & 0
\end{bmatrix} 
\begin{bmatrix}
0 \\ \cos\theta \\ \sin\theta
\end{bmatrix}
= \sigma_{yy}\cos^2\theta
\end{equation*}

Em particular, para certos valores de $\theta$:

\begin{equation*}
\theta = 30^\circ ~\therefore~ \sigma_n^{(\theta=30^\circ)} = \displaystyle\frac{3P}{4A}~;
\quad\quad
\theta = 45^\circ ~\therefore~ \sigma_n^{(\theta=45^\circ)} = \displaystyle\frac{P}{2A}
\end{equation*}

```python
print(' Tensao normal à faceta (30Deg) = ',
      '{:.2f}'.format(Syy*np.cos(np.deg2rad(theta))**2), 'MPa')


print(' Tensao normal à faceta (45Deg) = ',
      '{:.2f}'.format(Syy*np.cos(np.deg2rad(theta2))**2), 'MPa')


angrange = np.arange(0,91,5) # unit: degree

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(angrange,Syy*np.cos(np.deg2rad(angrange))**2, color='k', linestyle='-', linewidth=3.0, marker="o",
         markersize=10, label = r'$\theta$')
ax.set_xlabel(r'Angulo (deg)')
ax.set_ylabel(r'$\sigma_n = \sigma_{yy}\cos{^2}\theta$', color='k')
ax.grid()
```

A componente de $\vec{t}^{(\vec{n})}$ no plano da faceta, designa-se por tensão tangencial. O cálculo da tensão tangencial pode ser obtido atendendo ao teorema de Pitágoras:

\begin{equation*}
||\vec{t}^{(\vec{n})}||^2 = \sigma_n^2 + \tau^2
\end{equation*}

pelo que, o módulo da tensão tangencial vem,

\begin{equation*}
\tau = ||\vec{t}^{(\vec{n})} - \vec{\sigma}_n|| =
\sqrt{||\vec{t}^{(\vec{n})}||^2 - \sigma_n^2} =
\sqrt{\vec{t}^{(\vec{n})}\cdot\vec{t}^{(\vec{n})} - \sigma_n^2} =
\sqrt{t_x^2 + t_y^2 + t_z^2  - \sigma_n^2}
\end{equation*}

pelo que,

\begin{equation*}
\tau = \sqrt{\sigma_{yy}^2\cos^2\theta-\sigma_{yy}^2\cos^4\theta} =
\sqrt{\sigma_{yy}^2\cos^2\theta\left(1-\cos^2\theta\right)} =
\sqrt{\sigma_{yy}^2\cos^2\sin^2\theta} =
\sigma_{yy}\cos\theta\sin\theta
\end{equation*}

Em particular, para $\theta = 30^\circ$ tem-se:

\begin{equation*}
\tau(\theta=30^\circ) = \displaystyle\frac{\sqrt{3}P}{4A}
\end{equation*}

e, para $\theta = 45^\circ$:

\begin{equation*}
\tau(\theta=45^\circ) = \displaystyle\frac{P}{2A}
\end{equation*}

Note-se que:
\begin{equation*}
\begin{cases}
\theta = 0^\circ & \quad \sigma_n = \displaystyle\frac{P}{A}; \tau = 0 \\
\theta = 90^\circ & \quad \sigma_n = 0; \tau = 0 \\
\theta = 45^\circ & \quad \sigma_n = \displaystyle\frac{P}{2A}; \tau = \displaystyle\frac{P}{2A}
\end{cases}
\end{equation*}

```python
print(' Tensao tangencial à faceta (angulo 30 deg) = ',
      '{:.2f}'.format(Syy*np.cos(np.deg2rad(theta))*np.sin(np.deg2rad(theta))), 'MPa')

print(' Tensao tangencial à faceta (angulo 45 deg) = ',
      '{:.2f}'.format(Syy*np.cos(np.deg2rad(theta2))*np.sin(np.deg2rad(theta2))), 'MPa')

fig, ax1 = plt.subplots()
ax1.plot(angrange,Syy*np.cos(np.deg2rad(angrange))*np.sin(np.deg2rad(angrange)),
         color='k', linestyle='-', linewidth=3.0, marker="o", markersize=10)
ax1.set_xlabel(r'Angulo (deg)')
ax1.set_ylabel(r'$\tau = (P/A)\cos\theta\sin\theta$', color='k')
ax1.grid()

```
