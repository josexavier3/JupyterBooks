# I.1 Conceito de tensão. Vector tensão. Tensão normal e tangencial. Equações de equilíbrio e tensor das tensões.

## Problema 2 <a name="PI12"></a>

Para o estado de tensão referido a um sistema de coordenadas ortonormado $S(O,x, y,z)$ e caracterizado pelos vectores tensão abaixo indicados, obtenha o tensor das tensões, dando os valores adequados às constantes $a$, $b$ e $c$.

\begin{equation*}
\begin{cases}
\vec{t}{(p,\vec{\imath})} = 10 \vec{\imath}~+~a \vec{\jmath}~-~6\vec{k} \\
\vec{t}{(p,\vec{\jmath})} = 4 \vec{\imath}~+~7 \vec{\jmath}~+~4\vec{k} \\
\vec{t}{(p,\vec{k})} = b \vec{\imath}~+~c \vec{\jmath}~+~2\vec{k}
\end{cases}
\end{equation*}

## Resolução

O vetor de tensão que atua na faceta perpendicular ao eixo $x_i$ (com $i =$ 1, 2 e 3, eixos coordenados de um referencial ortonormado), com normal unitária exterior orientada segundo o sentido positivo desse eixo, $\vec{t}{(P,\vec{e}_i)}$, pode escrever-se por:

\begin{equation*}
\begin{cases}
\vec{t}{(p,\vec{e}_1)} = \sigma_{11} \vec{e}_1  + \sigma_{12} \vec{e}_2  + \sigma_{13} \vec{e}_3 \\
\vec{t}{(p,\vec{e}_2)} = \sigma_{21} \vec{e}_1  + \sigma_{22} \vec{e}_2  + \sigma_{23} \vec{e}_3 \\
\vec{t}{(p,\vec{e}_3)} = \sigma_{31} \vec{e}_1  + \sigma_{32} \vec{e}_2  + \sigma_{33} \vec{e}_3 \\
\end{cases}
\end{equation*}

em que, $\sigma_{ij}$ representa o elemento genérico da matriz do tensor das tensões como componente segundo o eixo $x_j$ do vetor das tensões atuante na faceta perpendicular ao eixo $x_i$ e cuja normal exterior é dirigida segundo o sentido positivo desse eixo.

A análise do equilíbrio de um volume elementar definido por três facetas normais a eixos coordenados em torno da partícula $p$, permite deduzir (pelo equilíbrio das forças igual a zero) a seguinte fórmula das tensões de Cauchy:

\begin{equation*}
\vec{t}{(p,\vec{n})} = \tilde{\sigma}^T(p) \cdot  \vec{n}
\end{equation*}

em que,

\begin{equation*}
\tilde{\sigma}^T(p) \equiv [\sigma]^T(p) =
\begin{bmatrix}
\sigma_{11} & \sigma_{22} & \sigma_{31} \\
\sigma_{12} & \sigma_{22} & \sigma_{32} \\
\sigma_{13} & \sigma_{23} & \sigma_{33}
\end{bmatrix} =
\begin{bmatrix}
10 & a & -6 \\
4 & 7 & 4 \\
b & c & 2
\end{bmatrix}
\end{equation*}

Pelo sumatório dos momentos em torno do ponto $p$ é possível ainda demostrar que o tensor das tensões $\tilde{\sigma}(p)$ é simétrico: $\tilde{\sigma}(p) = \tilde{\sigma}^T(p)$, ou seja (em notação indicial:

\begin{equation*}
\sigma_{ij} = \sigma_{ji}
\end{equation*}

Logo, resulta dque $a = 4$, $b=-6$ e $c = 4$.
