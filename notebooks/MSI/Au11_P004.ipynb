## Resolução

#### a) Qual deverá ser o diâmetro dos veios $AB$ e $CD$, para que a tensão admissível do aço não seja ultrapassada.<a name="A"></a>

([enunciatum](#PV114))

###### Equílibrio estático dos momentos de torção

A engrenagem traduz a transferência de rotação entre a roda motora (de maior diâmetro neste exemplo) e a roda movida (de menor diâmetro) numa relação de transmissão por multiplicação de 4:10 = 0,4. 

O momento de torção aplicado ao mecanismo na secção $B$ será equilibrado, na análise estática, por momentos de torção nos encastramentos nas secções em $A$ e $D$, bem como pelos momentos de torção exercidos pela força circunferencial gerada no contacto dos dentes das engrenagens $B$ e $C$.

O diagrama de corpo livre dos elementos do mecanismo permitem interpretar e escrever as equações de equilíbrio estático.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au11/Au11_P004_i2.jpg"
width="700">

- Veio $AB$:

\begin{equation*}
+\circlearrowleft~:~\sum M_x = 0
~\Leftrightarrow~
T_B - F\cdot r_B - T_A = 0
~\Leftrightarrow~
T_B = F\cdot r_B + T_A
\end{equation*}

- Veio $CD$:

\begin{equation*}
+\circlearrowright~:~\sum M_x = 0
~\Leftrightarrow~
T_D - F\cdot r_C  = 0
~\Leftrightarrow~
T_D = F\cdot r_C
\end{equation*}

Nestas duas expressões existem três incógnitas: $T_D, T_A, F$. O problema é estaticamente indeterminado. Uma equação adicional será necessária, neste caso obtida pela análise cinemática das engrenagens.

###### Cinemática das engrenagens

A condição cinemática das engrenagens estabelece:

\begin{equation*}
\begin{aligned}
\phi_B r_B = \phi_Cr_C
\end{aligned}
\end{equation*}

de onde se conclui a seguinte relação dos ângulos de rotação das engrenagens:

\begin{equation*}
\begin{aligned}
\phi_B  = \frac{r_C}{r_B} \phi_C = 2.5 \, \phi_C
\end{aligned}
\end{equation*}

Como as extremidades dos veios estão fixos (pela ligação ao exterior tipo encastramento):

\begin{equation*}
\begin{aligned}
\phi_{B/A}\equiv \phi_B &= \frac{T_{AB}L_{AB}}{JG} \quad\wedge\quad \phi_A = 0\\
\phi_{C/D}\equiv \phi_C &= \frac{T_{CD}L_{CD}}{JG} \quad\wedge\quad \phi_D = 0
\end{aligned}
\end{equation*}

De onde vem:

\begin{equation*}
\phi_B  = \frac{r_C}{r_B} \phi_C
~\Leftrightarrow~
\frac{T_{AB}L_{AB}}{JG} = \frac{r_C}{r_B} \frac{T_{CD}L_{CD}}{JG}
~\Leftrightarrow~
T_{AB} = \frac{r_C L_{CD}}{r_B L_{AB}} T_{CD}
\end{equation*}

De notar que a expressão do ângulo de rotação é dada genericamente por : $\phi = \int_A^B T(x)L(x)/J(x)G(x)\mathrm{d}x$. Nesta expressão o momento fletor deve ser definido considerando o diagrama de esforços do momento torsor, para um troço
genérico $AB$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au11/Au11_P004_i3.jpg"
width="700">

Resultando finalmente o seguinte sistema de equações,

\begin{equation*}
\begin{cases}
T_B &= F\cdot r_B + T_A \\
F &= \displaystyle\frac{T_D}{r_C} \\
T_A &= \displaystyle\frac{r_C L_{CD}}{r_B L_{AB}} T_D
\end{cases}
~\Leftrightarrow~
\begin{cases}
T_B &= \displaystyle\frac{r_B}{r_C}\left( \displaystyle\frac{L_{AB} + L_{CD}}{L_{AB}L_{CD}} \right) T_D\\
&- \\
&-
\end{cases}
\end{equation*}

```python
import numpy as np
from sympy.solvers import solve
from sympy import symbols

# dados
# ----------------------
# unidades: m. Pa, N
rB, rC = 100e-3, 40e-3
LAB, LCD = 0.4, 0.6
Sadm, G, TB = 205e6, 80e9, 6000 # unidades: N.m
# ----------------------

TD, TA, F = symbols('TD TA F')

sol1 = solve([TB - TA - F*rB,TD - F*rC,TD - (rB*LAB)/(rC*LCD)*TA],[TD, TA, F],dict=True)

TDn = sol1[0][TD]
print(f'TD = {TDn:.2f}  N.m')
TAn = sol1[0][TA]
print(f'TA = {TAn:.2f}  N.m')
Fn = sol1[0][F]
print(f' F = {Fn:.2f} N')
```

##### Fórmula de torção

Da fórmula de torção:

\begin{equation*}
\tau = \frac{T \rho}{J}
\end{equation*}

em que $\rho$ é a coordenada radial da posição do veio em relação ao seu centro, $J$ é o momento de inércia polar de área de secção tranversal; para uma secção massiça circular:

\begin{equation*}
J = \frac{\pi c^4}{2} \quad (\text{m}^4)
\end{equation*}

A tensão de corte máxima ocorre na periferia do veio:

\begin{equation*}
\tau_\mathrm{max} = \frac{T r}{J}
\end{equation*}

De notar que a equação de dimensionamento deve ser verificada para o momento torsor máximo. Neste mecanismo, este ocorre no veio $CD$: $T_\mathrm{max} = T_{CD} \equiv T_D$ (diagrama do momentos torsor é constante ao longo da barra $CD$).

Pelo que a equação de dimensionamento vem:

\begin{equation*}
\begin{aligned}
\tau_\mathrm{max} &\le  \tau_\mathrm{adm} \\
\frac{2T_D r}{\pi r^4} &\le \tau_\mathrm{adm} \\
\frac{2T_D}{\pi r^3} &\le \tau_\mathrm{adm} \\
r &\ge \left(\frac{2T_D}{\pi \tau_\mathrm{adm}}\right)^{1/3} \\
\end{aligned}
\end{equation*}

```python
r = symbols('r')

sol2 = solve([Sadm - 2*TDn/np.pi/r**3],[r],dict=True)
rn = sol2[0][r]
print(f'    raio mínimo = {rn*1e3:.2f} mm')
print(f'diâmetro mínimo = {2*rn*1e3:.2f} mm')

# ou:
#rn2 = (2*TDn/np.pi/Sadm)**(1/3)
#print('rn = ','{:.4f}'.format(rn2),' mm')
```

#### b) Para o diâmetro da alínea anterior, determine a correspondente rotação de torção da secção B, $\phi_B$.<a name="B"></a> ([enunciatum](#PV114))

\begin{equation*}
\begin{aligned}
\phi_B = \frac{T_{AB}L_{AB}}{J_{AB}G} \quad (T_{AB} \equiv T_A)
\end{aligned}
\end{equation*}

```python
J = np.pi*rn**4/2
print(f'J = {J:.4e} m4')
phiB = TAn*LAB/J/G
print(f'phiB = {phiB:.4e} radianos')
print(f'phiB = {np.rad2deg(float(phiB)):.3f} º')
```
