Sabendo que a secção D do veio CD e a secção E do veio EF estão ambas encastradas, e que é aplicado um momento de torção de 10 N.m na secção A determine:

[a)](#A) As tensões tangenciais máximas e mínimas que actuam na secção transversal dos três veios.

[b)](#B) O ângulo de rotação da secção A.

### Resolução

#### a) As tensões tangenciais máximas e mínimas que actuam na secção transversal dos três veios.


```python
import numpy as np

G = 77.e9
de, di = 25.e-3, 20.e-3
lAB, lEF, lCD = 1.6, 1.6, 1.2
rB, rC, rF = 30.e-3, 120.e-3, 40.e-3
TA = 10.

def Jp(ri,re):
    return np.pi*(re**4-ri**4)/2

J = Jp(di/2,de/2)
print(f'J = {J:.3e} m4')
```

**Diagrama de Corpo Livre**

_Troço AB_

\begin{equation*}
\sum M_\textrm{O} = 0
\quad\Leftrightarrow\quad
T_\textrm{A} = r_\textrm{B} F
\end{equation*}

_Troço CD_


\begin{equation*}
\sum M_\textrm{O} = 0
\quad\Leftrightarrow\quad
T_\textrm{D} - r_\textrm{C} F + r_\textrm{C} F^{'} = 0
\end{equation*}

_Troço EF_

\begin{equation*}
\sum M_\textrm{O} = 0
\quad\Leftrightarrow\quad
T_\textrm{E} = r_\textrm{F} F^{'}
\end{equation*}

Em resumo,

\begin{equation*}
\begin{cases}
T_\textrm{E} = r_\textrm{F} F^{'} \\
T_\textrm{D} + r_\textrm{C} F^{'} = r_\textrm{C} F \\
T_\textrm{E} = r_\textrm{F} F^{'}
\end{cases}
\end{equation*}

**Análise cinemática**

\begin{equation*}
r_\textrm{B} \phi_\textrm{B} = r_\textrm{C} \phi_\textrm{C} = r_\textrm{F} \phi_\textrm{F}
\end{equation*}

Onde,

\begin{equation*}
\phi_\textrm{F} = \displaystyle\frac{T_\textrm{E}L_\textrm{EF}}{JG}
\quad\wedge\quad
\phi_\textrm{C} = \displaystyle\frac{T_\textrm{D}L_\textrm{CD}}{JG}
\end{equation*}

De onde,

\begin{equation*}
\phi_\textrm{C} = \displaystyle\frac{r_\textrm{F}}{r_\textrm{C}} \phi_\textrm{F}
\end{equation*}

Resultando,

\begin{equation*}
T_\textrm{D} =
\left(\displaystyle\frac{r_\textrm{F}}{r_\textrm{C}}\right)
\left(\displaystyle\frac{L_\textrm{EF}}{L_\textrm{CD}}\right) T_\textrm{E}
\end{equation*}

\begin{equation*}
T_\textrm{E} =
\frac{\left(\displaystyle\frac{r_\textrm{C}}{r_\textrm{B}}\right)}
{\left(\displaystyle\frac{r_\textrm{F}}{r_\textrm{C}}\right)
 \left(\displaystyle\frac{L_\textrm{EF}}{L_\textrm{CD}}\right) +
 \left(\displaystyle\frac{r_\textrm{C}}{r_\textrm{F}}\right)}
T_\textrm{A}
\end{equation*}


```python
import sympy as sp

TE = ((rC/rB)/((rF/rC)*(lEF/lCD)+(rC/rF)))*TA
print(f'TE = {TE:.3f} N.m')

TD = (rF/rC)*(lEF/lCD)*TE
print(f'TD = {TD:.3f} N.m')
```

\begin{equation*}
\tau = \frac{T\rho}{J}
\end{equation*}

```python
def tau(T, J, r):
    return T*r/J

print('Veio AB: ---')
vAB_taui = tau(TA,J,di/2)
print(f'taui = {vAB_taui*1e-6:.3f} MPa')

vAB_taue = tau(TA,J,de/2)
print(f'taue = {vAB_taue*1e-6:.3f} MPa')
```

```python

print('Veio CD: ---')
vCD_taui = tau(TD,J,di/2)
print(f'taui = {vCD_taui*1e-6:.3f} MPa')

vCD_taue = tau(TD,J,de/2)
print(f'taue = {vCD_taue*1e-6:.3f} MPa')
```

```python

print('Veio EF: ---')
vEF_taui = tau(TE,J,di/2)
print(f'taui = {vEF_taui*1e-6:.3f} MPa')

vEF_taue = tau(TE,J,de/2)
print(f'taue = {vEF_taue*1e-6:.3f} MPa')
```

#### b) O ângulo de rotação da secção A.

\begin{equation*}
\phi_{A/B} = \phi_A-\phi_B = \displaystyle\frac{T_AL_{AB}}{JG}
\end{equation*}

\begin{equation*}
\phi_{B} = \displaystyle\frac{r_C}{r_B} \phi_C
\quad\wedge\quad
\phi_C = \displaystyle\frac{T_C L_{CD}}{JG}
\end{equation*}

De onde resulta,

\begin{equation*}
\phi_A  = \displaystyle\frac{T_AL_{AB}}{JG}
+ \left(\displaystyle\frac{r_C}{r_B}\right)\displaystyle\frac{T_C L_{CD}}{JG}
\end{equation*}


```python
phiA = (TA*lAB + (rC/rB)*TD*lCD)/J/G
print(f'phiA = {phiA:.4f} rad | {np.rad2deg(phiA):.2f} deg')
```
