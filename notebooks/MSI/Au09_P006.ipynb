##  Problema 6 <a name="PIV96"></a>

A figura representa uma barra de bronze AB colada a uma barra de alumínio BC. Sabendo que existe uma folga de 0,5 mm na extremidade esquerda da barra de bronze quando a temperatura é de 20º C, e que os apoios onde as extremidades das barras vão encostar são rígidos, determine:

[a)](#A) A temperatura para a qual a tensão normal na barra de alumínio é igual a -90 MPa.

[b)](#B) O correspondente comprimento exacto da barra de alumínio.

Dados: $A_\mathrm{bronze} = $ 1500 mm $^3$,
$E_\mathrm{bronze} = $ 105 GPa,
$\alpha_\mathrm{bronze} =
21.6\times10^{-6}$ ºC $^{-1}$,
$A_\mathrm{Alumínio} = $ 1800 mm $^3$,
$E_\mathrm{Alumínio} = $ 73 GPa,
$\alpha_\mathrm{Alumínio} =
23.2\times10^{-6}$ ºC $^{-1}$.

## Resolução

#### a) A temperatura para a qual a tensão normal na barra de alumínio é igual a -90 MPa.<a name="A"></a> ([_enunciatum_](#PIV96))

**Problema estaticamente indeterminado**

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au09/Au09_P006_i2.jpg"
width="650">

**Equação de compatibilidade de deslocamentos**

\begin{equation*}
|\delta^T| - |\delta'| = |\delta^F|
\end{equation*}

onde:

- $\delta'$ : folga
- $\delta^T$ : deslocamento nas barras de origem térmica
- $\delta^F$ : deslocamento nas barras de origem mecânica

**Relações força-deslocamento / expansão térmica-deslocamento**

Devido à composição em série da barra em barras de secção e material (bronze
e alumínio) diferente resulta:

\begin{equation*}
\delta^T = \delta_{b|_{\textrm{bronze}}}^T
+ \delta_{a|_{\textrm{aluminio}}}^T
= \alpha_b\,\Delta T\,L_{\textrm{AB}} + \alpha_a\,\Delta T\,L_{\textrm{BC}}
\end{equation*}

e

\begin{equation*}
\delta^F = \delta_{b|_{\textrm{bronze}}}^F
+ \delta_{a|_{\textrm{aluminio}}}^F
= \frac{F_b L_{\textrm{AB}}}{A_b E_b} + \frac{F_a L_{\textrm{BC}}}{A_a E_a}
\qquad\left(\textrm{barras compostas:}~\delta^F = \sum_i\frac{F_i\,L_i}{A_i\,E_i}\right)
\end{equation*}

em que $F_a$ e $F_b$ são o esforço axial em cada troço da barra AB (bronze) e  BC (alumínio).

**Equilíbrio estático**

Da análise do equilíbrio estático vem:

\begin{equation*}
\sum F = 0
\quad\Leftrightarrow\quad
F_A - F_C = 0
\quad\Leftrightarrow\quad
F_A = F_C = F
\end{equation*}

Pelo método das secções é fácil verificar o diagrama de esforços axial para dada troço AB e BC, este será constante e igual a $F$. O esforço axial será compressivo.

**Tensã normal**

\begin{equation*}
\sigma_a = \frac{F_a}{A_a} \equiv \frac{F}{A_a}
\quad\textrm{(compressão)}
\end{equation*}

**Sistema final de equações**

\begin{equation*}
\begin{aligned}
&\Delta T\left(\alpha_b\,L_{\textrm{AB}} + \alpha_a\,L_{\textrm{BC}}\right)
- \delta' = F\left(\frac{L_{\textrm{AB}}}{A_b E_b} + \frac{L_{\textrm{BC}}}{A_a
E_a}\right)\\
\Leftrightarrow\quad&
\Delta T\left(\alpha_b\,L_{\textrm{AB}} + \alpha_a\,L_{\textrm{BC}}\right)
- \delta' = \sigma_a A_a \left(\frac{L_{\textrm{AB}}}{A_b E_b} +
\frac{L_{\textrm{BC}}}{A_a
E_a}\right)
\end{aligned}
\end{equation*}

```python
import numpy as np

# units: mm, MPa, N, ºC

# Bronze (1)
Ab, Eb = 1500., 105.e3
coefDilatab = 21.6*1e-6

# Aluminio (2)
Aa, Ea = 1800., 73.e3
coefDilataa = 23.2*1e-6

dfolga = 0.5
Siga = -90.

LAB, LBC = 350., 450.

F = np.abs(Siga)*Aa
print(f' Força = {F:6.1f} N')

num = F*(LAB/Ab/Eb+LBC/Aa/Ea) + dfolga
dem = coefDilatab*LAB + coefDilataa*LBC
DT = num/dem
print(f' DT = {DT:6.1f} ºC')
```

#### b) O correspondente comprimento exacto da barra de alumínio.<a name="B"></a> ([_enunciatum_](#PIV96))

\begin{equation*}
\begin{aligned}
& \delta_a = \delta_a^T - \delta_a^F \\
\Leftrightarrow\quad &
\delta_a = \alpha_a \Delta T L_\textrm{BC} - \frac{F\,L_\textrm{BC}}{A_a\,E_a}
\end{aligned}
\end{equation*}

```python
daT = coefDilataa*DT*LBC
print(f' dT = {daT:6.3f} mm')
daF = F*LBC/Aa/Ea
print(f' dF = {daF:6.3f} mm')

da = daT - daF
print(f' da = dT - dF = {da:6.3f} mm')
```

Pelo que o comprimentos final será dado por:

\begin{equation*}
L_\textrm{Final} = L_\textrm{Inicial} + \delta_a
\end{equation*}

```python
LBCfinal = LBC + da
print(f' LBC final = {LBCfinal:6.3f} mm')
```
