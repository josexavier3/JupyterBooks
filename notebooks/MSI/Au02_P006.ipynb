# I.2 Transformação de tensões. Tensões principais. Direções principais.

## Problema 6 <a name="PI26"></a>

Considere o seguinte estado de tensão $\sigma_{ij}$

\begin{equation*}
\sigma_{ij} \equiv [\sigma_0] =
\begin{bmatrix}
3 & 0 & 6 \\
0  & 0 & 0 \\
6 & 0 & -3
\end{bmatrix}\quad\textrm{MPa}
\end{equation*}

Um sistema de eixos $S_1(0,x’,y’,z’)$ é obtido por uma rotação positiva de 120º, do referencial $S_0(0,x,y,z)$ em torno de um eixo $PQ$ que faz ângulos iguais com os 3 eixos $x,y,z$. Calcule o novo tensor das tensões se o tensor inicial for o indicado.



## Resolução

Considere-se dois sistemas de coordenadas distintos, $S_0(0,x,y,z)$ e $S_1(0,x',y',z')$. Os cosenos diretores dos ângulos entre os eixos coordenados $(x,y,z)$ e os eixos coordenados $(x',y',z')$ são dados pela seguinte tabela:

\begin{equation*}
\begin{array}{c|ccc}
\hline  & x & y & z \\\hline
  x' & l_1 & m_1 & n_1 \\
  y' & l_2 & m_2 & n_2 \\
  z' & l_3 & m_3 & n_3 \\\hline
\end{array}
\end{equation*}

Cada linha da matriz $(l,m,n)$ traduz o produto escalar entre os eixos coordenados $(x',y',z')$ no sistema de coordenadas $S_0(0,x,y,z)$. Para a primeira direção $x'$, têm-se portanto: $l_1=\cos(x',x), m_1=\cos(x',y), n_1=\cos(x',z)$. Por outro lado, cada coluna traduz o produto escalar entre os eixos coordenados $(x,y,z)$ no sistema de coordenadas $S_1(0,x',y',z')$. A título de exempmlo, para a direção $x$, têm-se: $l_1=\cos(x,x'), l_2=\cos(x,y'), l_3=\cos(x,z')$.

A matriz de tansformação do sistema $S_0$ para um novo referencial $S_1$ vem assim dada:

\begin{equation*}
[T_{01}] =
\begin{bmatrix}
l_1 & m_1 & n_1 \\
l_2 & m_2 & n_2 \\
l_3 & m_3 & n_3
\end{bmatrix} =
\begin{bmatrix}
\cos(x',x) & \cos(x',y) & \cos(x',z) \\
\cos(y',x) & \cos(y',y) & \cos(y',z) \\
\cos(z',x) & \cos(z',y) & \cos(z',z)
\end{bmatrix}
\end{equation*}

Observando segundo a direção octaédrica (vetor $\vec{PQ}$), após rotação de 120$^\circ$, verifica-se que a direção inicial do eixo coordenado $x$ passa a coincidir com o eixo coordenado $z'$, o eixo coordenado $y$ passa a coincidir com o eixo coordenado $x'$ e o eixo coordenado $z$ passa a coincidir com o eixo coordenado $y'$. Isto é:

\begin{equation*}
\begin{cases}
\vec{\imath}' &= ~\vec{\jmath} \\
\vec{\jmath}' &= ~\vec{k} \\
\vec{k}' &= ~\vec{\imath}
\end{cases}
\end{equation*}

ou seja,

\begin{equation*}
[T_{01}] =
\begin{bmatrix}
0 & 1 & 0 \\
0  & 0 & 1 \\
1 & 0 & 0
\end{bmatrix}
\end{equation*}

De acordo com a lei de transformação de um tensor de segunda ordem simétrico vem:

\begin{equation*}
[\sigma_1] = [T_{01}][\sigma_0][T_{01}]^T
\end{equation*}

algebricamente,

\begin{equation*}
\sigma_{ij}' =
\begin{bmatrix}
0 & 1 & 0 \\
0  & 0 & 1 \\
1 & 0 & 0
\end{bmatrix}
\begin{bmatrix}
3 & 0 & 6 \\
0  & 0 & 0 \\
6 & 0 & -3
\end{bmatrix}
\begin{bmatrix}
0 & 0 & 1 \\
1  & 0 & 0 \\
0 & 1 & 0
\end{bmatrix} =
\begin{bmatrix}
0 & 0 & 0 \\
0 & -3 & 6 \\
0 & 6 & 3
\end{bmatrix}~\textrm{(MPa)}
\end{equation*}
