### Resolução

#### a) Calcule o vetor tensão aplicado num plano perpendicular ao versor $\vec{n} \equiv \hat{n} = (0,11; 0,35; 0,93)$ e que passa por $P$.


O princípio das tensões de Cauchy estabelece que as interações materiais que ocorrem em superfícies internas de um corpo
podem ser descritas como uma distribuição de forças de tração, da mesma forma como é descrito o efeito de forças externas
que atuam nas superfícies físicas do corpo.

A partir da representação matricial do tensor das tensões num determinado ponto $P$,
$(\tilde{\sigma} \equiv \underline{\underline{\sigma}})$,
é possível determinar o vetor tensão, $\vec{t}^{(P,\vec{n})}$, que atua nesse ponto e sobre uma faceta  cuja orientação no espaço
é definida pelo versor exterior à superfície, $\vec{n}$. Essa relação é conhecida pela
fórmula da tensão de Cauchy, sendo expressa por:

\begin{equation}
\begin{aligned}
\vec{t}^{(P,\vec{n})} &= \tilde\sigma^T(P) \cdot \vec{n}
\quad|\quad \underline{t} (P,\vec{n}) = \underline{\underline{\sigma}}^T(P) \cdot \underline{n}
&\quad\mbox{(notação direta)} \\
t_i &= \sigma_{ji} \cdot n_j &\quad\mbox{(notação indicial)}
\end{aligned}
\end{equation}

Num sistema de coordenadas $S(0,\vec{e}_1,\vec{e}_2,\vec{e}_3)$ e para efeitos de cálculo pode usar-se a seguinte notação:

\begin{equation}
\begin{aligned}
\begin{bmatrix}
t_1 \\ t_2 \\ t_3
\end{bmatrix} &=
\begin{bmatrix}
\sigma_{11} & \sigma_{21} & \sigma_{31} \\
\sigma_{12} & \sigma_{22} & \sigma_{32} \\
\sigma_{13} & \sigma_{23} & \sigma_{23}
\end{bmatrix}
\begin{bmatrix}
n_1 \\ n_2 \\ n_3
\end{bmatrix}
\end{aligned}
\end{equation}

em que, cada coluna da matriz $\underline{\underline{\sigma}}^T(P)$ é obtida pelas componentes cartesianas dos vetores
tensão, $\vec{t}(\vec{e}_j)$, que atuam nas facetas normais aos eixos coordenados do sistema ortonormado e direto:

\begin{equation}
\begin{cases}
\vec{t}(\vec{e}_1) = \sigma_{11} \vec{e}_1 + \sigma_{12} \vec{e}_2 + \sigma_{13} \vec{e}_3\\[1ex]
\vec{t}(\vec{e}_2) = \sigma_{21} \vec{e}_1 + \sigma_{22} \vec{e}_2 + \sigma_{23} \vec{e}_3\\[1ex]
\vec{t}(\vec{e}_3) = \sigma_{31} \vec{e}_1 + \sigma_{32} \vec{e}_2 + \sigma_{33} \vec{e}_3\\
\end{cases}
\end{equation}

Em alternativa, o cálculo das componentes do vetor tensão num ponto e numa determinada faceta pode ser obtido por:

\begin{equation}
\begin{bmatrix}
t_1 & t_2 & t_3
\end{bmatrix} =
\begin{bmatrix}
n_1 & n_2 & n_3
\end{bmatrix}
\begin{bmatrix}
\sigma_{11} & \sigma_{12} & \sigma_{13} \\
\sigma_{21}(=\sigma_{12}) & \sigma_{22} & \sigma_{23} \\
\sigma_{31}(=\sigma_{13}) & \sigma_{32}(=\sigma_{23}) & \sigma_{23}
\end{bmatrix}
\end{equation}

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def versor(n1,n2,n3):
    return np.array([n1, n2, n3])

def TensorTensoes(s11,s12,s13,s21,s22,s23,s31,s32,s33):
    return np.array([[s11,s12,s13], [s21,s22,s23], [s31,s32,s33]])

# visualizacao do versor n:
print("visualizacao da orientação espacial do versor n:")
soa = np.array([[0, 0, 0, 0.11, 0.35, 0.93]])
X, Y, Z, U, V, W = zip(*soa)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([0, .11])
ax.set_ylim([0, .35])
ax.set_zlim([0, .6])
ax.view_init(35, 65)
plt.show()

versor_n = versor(0.11, 0.35, 0.93)
Sigma_n = TensorTensoes(1000, 200, 0, 200, -6000, -400, 0, -400, 0)
vetor_tensao_n = np.dot(versor_n,Sigma_n)

print("O vetor tensão no ponto P num plano perpendicular ao versor (0,11; 0,35; 0,93) é :")
print("t^n = ",vetor_tensao_n, "kPa")
```

#### b) Calcule as componentes de tensão normal e tangencial a esse plano.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au01/P001/Au01_P001_i1.jpg" width="250" height="200">

À componente do vetor das tensões, $\vec{t}^{(\hat{n})}\equiv \vec{\sigma}^{(\vec{n})}$. segundo a normal à faceta
designa-se por tensão normal, $\sigma_n$, e pode ser obtida fazendo o produto interno
de $\vec{t}^{(\vec{n})}$ por $\vec{n}$:

\begin{equation}
\sigma_n(p,\vec{n}) = \vec{t}^{(\vec{n})} \cdot \vec{n} = t_i^{(\vec{n})} \cdot n_i
\quad \mbox{;} \quad
\vec{\sigma}_n(p,\vec{n}) = (\vec{t}^{(\vec{n})} \cdot \vec{n})\vec{n}
\end{equation}

A tensão normal diz-se de tração se for positiva e diz-se de compressão no caso negativo.


```python
tensao_normal = np.dot(vetor_tensao_n,versor_n)
print("A componente de tensão normal é: ",'{:.2f}'.format(tensao_normal), "kPa")
```


A componente de $\vec{t}^{(\vec{n})}$ no plano da faceta, designa-se por tensão tangencial.
O cálculo da tensão tangencial pode ser obtido atendendo ao teorema de Pitágoras:

\begin{equation}
||t_i^{(\vec{n})}||^2 = \sigma_n^2 + \tau^2
\end{equation}

pelo que, o módulo da tensão tangencial vem,

\begin{equation}
\tau(p,\vec{n}) = ||\vec{t}^{(\vec{n})} - \vec{\sigma}_n|| =
\sqrt{\vec{t}^{(\vec{n})}\cdot\vec{t}^{(\vec{n})} - \sigma_n^2} \quad \mbox{;} \quad
\vec{\tau}(p,\vec{n}) \equiv \tau\vec{s} = \vec{t} - (\vec{t}^{(\vec{n})} \cdot \vec{n})\vec{n}
\end{equation}


```python
vt_n2 = np.dot(vetor_tensao_n,vetor_tensao_n)
tensao_tangencial = np.sqrt(vt_n2 - tensao_normal**2)
print("A componente da tensão tangencial é: ",'{:.2f}'.format(tensao_tangencial), "kPa")
```
