#### a) As tensões normal e tangencial ao plano diagonal indicado.<a name="A"></a>

([_enunciatum_](#PI31))

Considere-se um elemento de volume infinitesimal circunscrito no ponto material ou partícula $P$ tal que: $\mathrm{d}x = \mathrm{d}z$ e $\mathrm{d}y = 2\mathrm{d}x$ (ver figura).

De acordo com o estado de tensão representado na figura, existe um estado plano de tensão com tensões não nulas definidas no plano cartesiano $y-z$.
Considere-se agora o seguinte plano inclinado cuja normal é dada pelo versor $\vec{n}$ :

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au03/Au03_P001_i2.jpg"
width="300" height="250">

O ângulo $\theta$ (ou, de forma equivalente o ângulo $\alpha$) carateriza completamente a transformação do sistema de coordenadas $S_0(0,\vec{\imath},\vec{\jmath})$ para o novo referencial $S_1(0,\vec{n},\vec{t})$ (em que $\vec{n}$ é normal à faceta considerada). O amplitude do ângulo $\theta$ pode ser obtido algebricamente considerando a seguinte relação trigonométrica:

\begin{equation*}
\tan\theta = \frac{\mathrm{d}y}{\mathrm{d}z} =
2 \quad \Leftrightarrow\quad \theta = \tan^{-1}(2)
\end{equation*}

De forma equivalente o ângulo $\alpha$ vem:

\begin{equation*}
\tan\alpha = \frac{\mathrm{d}z}{\mathrm{d}y}
= \frac{1}{2} \quad \Leftrightarrow\quad \alpha = \tan^{-1}(1/2)
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 18
# rc('text', usetex=True)

def atan(r): return np.arctan(r), np.rad2deg(np.arctan(r))
# see also: np.arctan2(adjacent,opposite)

dydz = 2
theta_rad, theta_deg = atan(dydz)

print('    r :  dydz = ','{:.3f}'.format(dydz))
print('angulo: THETA = ','{:.3f}'.format(theta_rad), ' rad | ','{:.3f}'.format(theta_deg),' deg')

dzdy = 1/2
alpha_rad, alpha_deg = atan(dzdy)

print('    r :  dzdy = ','{:.3f}'.format(dzdy))
print('angulo: ALPHA = ','{:.3f}'.format(alpha_rad), ' rad | ','{:.3f}'.format(alpha_deg),' deg')
print('THETA + ALPHA =','{:.3f}'.format(theta_rad+alpha_rad),' rad | ','{:.3f}'.format(theta_deg+alpha_deg),' deg')
print('(r = ratio of the side lengths opposite to the angle and adjacent the angle)')

font = {'size' : 15}
plt.rc('font', **font)
plt.rcParams['axes.facecolor']='black'
a = np.linspace(-np.pi, np.pi)
plt.plot(a, np.arctan(a))
ax = plt.gca()
plt.plot(dydz, theta_rad, 'ro', markersize=12)
plt.plot(dzdy, alpha_rad, 'gs', markersize=12)
ax.axhline(y=0, color='gray'), ax.axvline(x=0, color='gray')
plt.text(1.0*dydz, .7*theta_rad, r'$\theta$', color='w', fontsize='xx-large')
plt.text(1.5*dzdy, .7*alpha_rad, r'$\alpha$', color='w', fontsize='xx-large')
plt.xlabel(r"$r$ = opposite/adjacent", size=18)
plt.ylabel(r"$\tan^{-1}\theta$", size=18)
plt.axis('tight')
plt.grid('on')
plt.show()
```

Considere-se a expressão do versor $\vec{n}$ em função do ângulo $\theta$:

\begin{equation*}
\begin{aligned}
\vec{n} = \underline{n} &=
\begin{bmatrix}%
\cos(\vec{n},\vec{\imath}) &
\cos(\vec{n},\vec{\jmath}) &
\cos(\vec{n},\vec{k})
\end{bmatrix}%
\\
& =
\begin{bmatrix}%
\cos(\pi/2) & \cos(\theta) & \cos(\pi/2-\theta)
\end{bmatrix}%
\\
& =
\begin{bmatrix}%
0 & \cos\theta & \sin\theta
\end{bmatrix}%
\end{aligned}%
\end{equation*}

```python
def versor(e1,e2,e3): return np.array([e1,e2,e3])

lt = np.around(np.cos(np.pi/2),decimals=6)
mt = np.cos(theta_rad)
nt = np.cos(np.pi/2-theta_rad)
n1t = versor(lt,mt,nt)
print('\nn (theta) = ',n1t)
```

Ou, por outro lado, considerando a expressão do versor $\vec{n}$ em função do
ângulo $\alpha$:

\begin{equation*}
\begin{aligned}
\vec{n} = \underline{n} &=
\begin{bmatrix}%
\cos(\vec{n},\vec{\imath}) &
\cos(\vec{n},\vec{\jmath}) &
\cos(\vec{n},\vec{k})
\end{bmatrix}%
\\
& =
\begin{bmatrix}%
\cos(\pi/2) & \cos(\pi/2-\alpha) & \cos(\alpha)
\end{bmatrix}%
\\
& =
\begin{bmatrix}%
0 & \sin\alpha & \cos\alpha
\end{bmatrix}%
\end{aligned}%
\end{equation*}

```python
la = np.around(np.cos(np.pi/2),decimals=6)
ma = np.cos(np.pi/2-alpha_rad)
na = np.cos(alpha_rad)
n1a = versor(la,ma,na)
print('\nn (alpha) = ',n1a)
```

O vetor de tensão pode ser obtida pela fórmula das tensões de Cauchy. Em notação direta escreve-se:

\begin{equation*}
\vec{t}(\vec{x},\vec{n},t) = \tilde{\sigma}^{\,T}(\vec{x}, t) \cdot \vec{n}
\end{equation*}

em notação matricial têm-se,

\begin{equation*}
\underline{t}(\vec{x},\vec{n},t) =
\underline{\underline{\sigma}}^{\,T}(\vec{x}, t) \underline{n}
\end{equation*}

desenvolvendo,

\begin{equation*}
\begin{aligned}
\begin{bmatrix}
t_x \\ t_y \\ t_z
\end{bmatrix}
& =
\begin{bmatrix}
\sigma_{xx} & \tau_{yx} & \tau_{zx} \\
\tau_{xy} & \sigma_{yy} & \tau_{zy} \\
\tau_{xz} & \tau_{yz} & \sigma_{zz}
\end{bmatrix}
\begin{bmatrix}
n_x \\ n_y \\ n_z
\end{bmatrix}  \\
&=
\begin{bmatrix}
\vec{t}(\vec{\imath}) & \vec{t}(\vec{\jmath}) & \vec{t}(\vec{k})
\end{bmatrix}
\begin{bmatrix}
n_x \\ n_y \\ n_z
\end{bmatrix}
\end{aligned}
\end{equation*}

De notar que a primeira coluna da matriz $\tilde{\sigma}^{\,T}$ é formada pelas
componentes da tensão $\vec{t}(\vec{\imath})$, e assim sucessivamente, como
ilustra a figura em baixo. As tensões em três facetas de um sistema de
coordenadas cartesiano vêm assim defindas:

\begin{equation*}
\begin{aligned}
\vec{t}(\vec{\imath}) &= \sigma_{xx} \vec{\imath} + \tau_{yx}\vec{\jmath} +
\tau_{zx} \vec{k} \\
\vec{t}(\vec{\jmath}) &= \tau_{xy} \vec{\imath} + \sigma_{yy}\vec{\jmath} +
\tau_{zy} \vec{k} \\
\vec{t}(\vec{k}) &= \tau_{xz} \vec{\imath} + \tau_{yz}\vec{\jmath} +
\sigma_{zz} \vec{k}
\end{aligned}
\end{equation*}

Em notação indicial a fórmula de Cauchy vem:

\begin{equation*}
t_i = \sigma_{ji}n_j
\end{equation*}

O tensor $\tilde{\sigma}^{\,T}(\vec{x}, t)$ representa uma aplicação linear que num dado ponto material $P$ dum sólido,
associa a cada versor $\vec{n}$ o vetor tensão $\vec{t} (\vec{x},\vec{n})$, que atua numa faceta que contém
a partícula $P$ e que é ortogonal a $\vec{n}$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au03/Au03_P001_i4.jpg" width="400" height="190">

A componente do vetor das tensões, $\vec{t}^{(\vec{n})}$, segundo a faceta com normal $\vec{n}$,
designa-se por tensão normal, $\sigma_n$, e pode ser obtida pelo produto escalar
entre os vetores $\vec{t}^{(\vec{n})}$ e $\vec{n}$:

\begin{equation*}
\begin{aligned}
\sigma_n(P,\vec{n}) &= \vec{t}^{(\vec{n})} \cdot \vec{n}
= t_i^{(\vec{n})} n_i = n_i \sigma_{ji} n_j =
\underline{n}^\mathrm{T}\,\underline{\underline{\sigma}}\,\underline{n}\\
\vec{\sigma}_n(P,\vec{n}) &= (\vec{t}^{(\vec{n})} \cdot \vec{n}) \vec{n}
\end{aligned}
\end{equation*}

A tensão normal diz-se de tração se for positiva e diz-se de compressão no caso negativo.

A componente de $\vec{t}^{(\vec{n})}$ no plano da faceta, designa-se por tensão de corte.
O seu cálculo pode ser obtido atendendo ao teorema de Pitágoras:

\begin{equation*}
||t_i^{(\vec{n})}||^2 = \sigma_n^2 + \tau^2
\end{equation*}

pelo que, o módulo da tensão de corte vem,

\begin{equation*}
\tau(P,\vec{n}) = ||\vec{t}^{(\vec{n})} - \vec{\sigma}_n|| =
\sqrt{\vec{t}^{(\vec{n})}\cdot\vec{t}^{(\vec{n})} - \sigma_n^2} \quad \textrm{;} \quad
\vec{\tau}(P,\vec{n}) = \vec{t} - (\vec{t}^{(\vec{n})} \cdot \vec{n})\vec{n}
\end{equation*}

sendo o vetor de tensão de corte dado por:

\begin{equation*}
\vec{\tau}(P,\vec{n}) = \tau(P,\vec{n}) \vec{s}
\end{equation*}

em que $\vec{s}$ representa o versor normal a $\vec{n}$ coplanar à faceta no ponto $P$.

```python
np.set_printoptions(precision=3)

def sij(sx,txy,txz,sy,tyz,sz):
    return np.array([[sx,txy,txz],[txy,sy,tyz],[txz,tyz,sz]])

sij_xyz = sij(0,0,0,-500,80,100)
print('\nsij_xyz = \n',sij_xyz)
print('\nn1 = ',n1a)
t1 = np.dot(n1a,sij_xyz)
print('t1 = ',t1, 'MPa')

sn = np.dot(t1,n1a)
print('\nsn = ','{:.1f}'.format(sn),' MPa')

tau = np.sqrt(np.dot(t1,t1) - sn**2)
print('tau = ','{:.1f}'.format(tau),' MPa')
```

#### b) As tensões principais.<a name="B"></a>

([enunciatum](#PI31))

A determinação das tensões principais pode ser obtida por diferentes abordagens:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au03/detsij.jpg" width="400" height="210">

Recorrendo à solução pelo determinante, as tensões principais podem obtêm-se pela expressão:

\begin{equation*}
\det(\sigma_{ij}-\sigma\delta_{ij})= 0
\end{equation*}

A resolução algebrica desta equação conduz ao seguinte polinómio caraterístico (cúbico) associado ao tensor das tensões:

\begin{equation*}
\sigma^3 - I_1\sigma^2 + I_2\sigma - I_3 = 0
\end{equation*}

em que $I_i$ ($i=1,2,3$) representam os invariantes do tensor das tensões. Os invariantes vêm expressos em função das componentes cartesianas da matriz de tensão, $\sigma_{ij}$:

\begin{equation*}
\begin{aligned}
I_1 &= \mathrm{Tr}(\sigma_{ij}) \\
I_2 &= \frac{1}{2}\left[ \mathrm{Tr}(\sigma_{ij})^2 - \mathrm{Tr}(\sigma_{ij}^2) \right] =
\frac{1}{2}\left( \sigma_{ii}\sigma_{jj} - \sigma_{ij}\sigma_{ji} \right) \\
& =
\begin{vmatrix}
\sigma_{11} & \sigma_{12} \\
\sigma_{21} & \sigma_{22} \\
\end{vmatrix} +
\begin{vmatrix}
\sigma_{11} & \sigma_{13} \\
\sigma_{31} & \sigma_{33} \\
\end{vmatrix} +
\begin{vmatrix}
\sigma_{22} & \sigma_{23} \\
\sigma_{32} & \sigma_{33} \\
\end{vmatrix} \\\\
I_3 &= \mathrm{det}(\sigma_{ij}) \\
\end{aligned}
\end{equation*}

```python
print('--------------------------------------------------')
print('Solução pelo polinómio caraterístico (invariantes)')
print('--------------------------------------------------')

tt0 = sij_xyz

I1 = np.trace(tt0)
print('\n1º Invariante, I1:', '{:.1f}'.format(I1))

I2 = 0.5*(np.trace(tt0)**2 -np.trace(np.dot(tt0,tt0)))
print('2º Invariante, I2 :', '{:.1f}'.format(I2))

I3 = np.linalg.det(tt0)
print('3º Invariante, I3 :', '{:.1f}'.format(I3))

print('\nDeterminação numérica das raizes do polinómio - Numpy :')
valpropriosN = np.roots([1,-I1,I2,-I3])
print('\n Numpy : roots = ',valpropriosN)
valpropriosN = np.sort(valpropriosN)
sIII = valpropriosN[0]
sII = valpropriosN[1]
sI = valpropriosN[2]
print('\n  Tensão principal (valor próprio) sI = ','{:.2f}'.format(sI))
print(' Tensão principal (valor próprio) sII = ','{:.2f}'.format(sII))
print('Tensão principal (valor próprio) sIII =','{:.2f}'.format(sIII))

srange = np.linspace(sIII*1.2,sI*1.2,50)
def polycaract(s):
    return s**3 - I1*s**2 + I2*s - I3

font = {'size' : 17}
plt.rc('font', **font)
plt.rcParams['axes.facecolor']='black'
plt.figure(figsize=(9,6))
plt.plot(srange,polycaract(srange),linewidth=3)
for i in valpropriosN:
    plt.plot(i,polycaract(i),'o',markersize=20)
plt.title('Polinómio característico')
plt.ylabel('$f(\sigma)$: poly3')
plt.xlabel('$\sigma$')
sleg = ('I','II','III')
j = 0
for i in (sI,sII,sIII):
    plt.text(i,-.5e7,'$\sigma_\mathrm{'+sleg[j]+'}$',color='white',fontfamily='monospace',fontsize='xx-large')
    j+=1
plt.grid('on')
plt.show()
```

#### c) As tensões principais. Pelo método da circunferência de Mohr.<a name="B"></a>

([enunciatum](#PI31))

A circunferência de Mohr é uma representação gráfica da transformação do estado de tensão num ponto
introduzida pelo Eng. Alemão Otto Mohr (1835-1918). Esta representação gráfica pode iniciar-se pelo conhecimento das
componentes do tensor das tensões num determinado sistema de coordenadas. Considere-se o plano $(\sigma,\tau)$.
A partir das posições de dois pontos, $A$ e $B$, correspondentes às tensões normal e tangencial em duas
facetas ortogonais (estado plano de tensão), é possível definir um círculo cujo diâmetro é definido pelo segmento de
reta $\overline{AB}$. Desta figura geométrica
podem extraem-se duas dimensões: (i) a posição do centro da circunferência, $C$;
(ii) o valor do raio da circunferência, $R$. Das relações geométricas têm-se, por exemplo,
no plano de análise $y-z$:

\begin{equation*}
\begin{aligned}
C &= \frac{\sigma_{yy} + \sigma_{zz}}{2} \\
R &= \sqrt{\left(\frac{\sigma_{xx}-\sigma_{zz}}{2}\right)^2 +
\tau_{yz}^2 } \\
\end{aligned}
\end{equation*}

Neste curso, a contrução da circunferência de Mohr, associada a um tensor de segunda ordem simétrico
bidimensional $[\sigma]$, segue as seguintes convenções:

1. A circunferência é desenhada num plano utilizando um sistema de eixos em que as abcissas representam as componentes
diagonais da matriz das tensões ($\sigma_{ii}$) e cujas ordenadas representam componentes não diagonais, ou seja, de
tensões de corte ($\tau_{ij}$).

2. As componentes do tensor bidimensional ($[\sigma] = [\sigma_{xx} ~ \tau_{xy}; \tau_{yx} ~ \sigma_{yy}]$) num referencial
ortonormado são as coordenadas de um par de pontos diametralmente opostos da circunferência de Mohr, de acordo com a
convenção:

2.1 Se a tensão de corte no plano em análise indicar o sentido anti-horário ($-\tau$), então:

\begin{equation*}
\begin{aligned}
& \textrm{Ponto A (abcissa, ordenada):}~(\sigma_{xx}, -\tau_{xy})\\
& \textrm{Ponto B (abcissa, ordenada):}~(\sigma_{yy}, \tau_{yx})
\end{aligned}
\end{equation*}

2.2 Se o sentido de rotação das tensões de corte for horário ($\tau$), então:

\begin{equation*}
\begin{aligned}
& \textrm{Ponto A (abcissa, ordenada):}~(\sigma_{xx}, \tau_{xy})\\
& \textrm{Ponto B (abcissa, ordenada):}~(\sigma_{yy}, -\tau_{yx})
\end{aligned}
\end{equation*}

3. Os ângulos na circunferência de Mohr são marcados com valor duplo dos da realidade.

4. Os ângulos na circunferência de Mohr são marcados no mesmo sentido ao do espaço físico.

As tensões principais $(\sigma_{\mathrm{I}}, \sigma_{\mathrm{II}})$ são obtidas diretamente pelos pontos de interseção
da circunferência com o eixo das abcissas (estes pontos não têm ordenadas uma vez que nos planos principais as tensões
de corte são nulas por definição).

Conclui-se, portanto, que a análise da circunferência de Mohr permite: (i) determinar os valores e direções principais
de um tensor conhecidas as suas componentes num referêncial dado; (ii) determinar as componentes do tensor das tensões
relativas a um referencial arbitrário conhecidas as suas componentes num determinado sistema de coordenadas.

Note-se que a circunferência de Mohr permite concluir que é no referencial principal que as componentes diagonais do
tensor das tensões, i.e., das tensões normais, têm valores extremos (máximo e mínimo), sendo estes iguais aos valores
principais.

Por outro lado, a componente $\tau_{xy}$ tem valor absoluto máximo igual a $R$ (raio da circunferência) e os eixos
correspondentes a tais componentes formam um ângulo de 45$^\circ$ relativamente aos eixos do referencial principal
(ou seja, um ângulo de 90$^\circ$ no espaço das tensões, $\sigma-\tau$, da circunferência de Mohr).

Finalmente, se as duas componentes principais do tensor forem iguais: $\sigma_{\mathrm{I}} = \sigma_{\mathrm{II}}$,
(tensor isotrópico), os pontos $\mathrm{I}$ e $\mathrm{II}$ coincidem e a circunferência de Mohr reduz-se a um ponto. Isto significa que
as componentes do tensor não variam numa transformação ortogonal de coordenadas.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au03/Au03_P001_i5.jpg" width="550" height="380">

Em alternativa ao círculo de Mohr, é possível analisar a transformação do tensor das tensões diretamente através das
expressões analíticas. Considerando a lei de transformação de um tensor de segunda ordem em estado plano de tensão:

\begin{equation*}
\begin{aligned}
[\sigma]_1 &= [T_{01}]\,[\sigma]_0\,[T_{01}]^{\mathrm{T}}
\end{aligned}
\end{equation*}

cujo desenvolvimento algébrico conduz a,

\begin{equation*}
\begin{aligned}
\sigma_{y'y'} &= \frac{\sigma_{yy} + \sigma_{zz} }{2} + \frac{\sigma_{yy} - \sigma_{zz} }{2} \cos2\theta -
\tau_{yz}\sin2\theta \\
\sigma_{z'z'} &= \frac{\sigma_{yy} + \sigma_{zz} }{2} - \frac{\sigma_{yy} - \sigma_{zz} }{2} \cos2\theta -
\tau_{yz}\sin2\theta \\
\tau_{y'z'} &= \frac{\sigma_{zz} - \sigma_{yy} }{2} \sin2\theta + \tau_{yz}\cos2\theta \\
\end{aligned}
\end{equation*}

A análise do círculo de Mohr permite determinar os valores máximos e mínimos das componentes normais das tensões num determinado plano de análise. Considere-se esse plano $Oyz$, vem então:

\begin{equation*}
\begin{aligned}
\sigma_\mathrm{max/min} \equiv \sigma_\mathrm{I/III} &= \sigma_\mathrm{médio} \pm R \\
&= \frac{\sigma_{yy} + \sigma_{zz}}{2} \pm  \sqrt{\left(\frac{\sigma_{yy}-\sigma_{zz}}{2}\right)^2 +
\tau_{yz}^2 }
\end{aligned}
\end{equation*}

```python
print('--------------------------')
print('Equações de transformação:')
print('--------------------------')

print('\nMatriz das tensões tt0 = \n',tt0)

def sxx1(sxx,syy,txy,t):
    return (sxx+syy)/2 + (sxx-syy)/2*np.cos(2*t) + txy*np.sin(2*t)

def syy1(sxx,syy,txy,t):
    return (sxx+syy)/2 - (sxx-syy)/2*np.cos(2*t) - txy*np.sin(2*t)

def txy1(sxx,syy,txy,t):
    return (syy-sxx)/2*np.sin(2*t) + txy*np.cos(2*t)

def angPrincipal(sxx,syy,txy):
    return np.arctan(2*txy/(sxx-syy))/2

def angS(s11,s22,s12): return -np.arctan((s11-s22)/(2*s12))

print('\nPLANO y-z:\n')
sxx = tt0[1,1] # neste caso: x = y; y = z
txy = tt0[1,2]
syy = tt0[2,2]

angt = np.arange(-45,145,1)
tsxx = sxx1(sxx,syy,txy,np.deg2rad(angt))
tsyy = syy1(sxx,syy,txy,np.deg2rad(angt))
ttxy = txy1(sxx,syy,txy,np.deg2rad(angt))

print('Tensões (normais) principais: -------------------------------------------')
aprinc = np.rad2deg(angPrincipal(sxx,syy,txy))
print('angPrincipal 1 (deg) =','{:.2f}'.format(aprinc))
tsxx1 = sxx1(sxx,syy,txy,np.deg2rad(aprinc))
tsyy1 = syy1(sxx,syy,txy,np.deg2rad(aprinc))
ttxy1 = txy1(sxx,syy,txy,np.deg2rad(aprinc))
print('tens xx plano principal [1] (MPa) =','{:.2f}'.format(tsxx1))
print('tens yy plano principal [1] (MPa) =','{:.2f}'.format(tsyy1))
print('tens xy plano principal [1] (MPa) =','{:.2f}'.format(ttxy1))

aprinc2 = aprinc + 90
print('\nangPrincipal 2 (deg) =','{:.2f}'.format(aprinc2))
tsxx2 = sxx1(sxx,syy,txy,np.deg2rad(aprinc2))
tsyy2 = syy1(sxx,syy,txy,np.deg2rad(aprinc2))
ttxy2 = txy1(sxx,syy,txy,np.deg2rad(aprinc2))
print('tens xx plano principal [2] (MPa) =','{:.2f}'.format(tsxx2))
print('tens yy plano principal [2] (MPa) =','{:.2f}'.format(tsyy2))
print('tens xy plano principal [2] (MPa) =','{:.2f}'.format(ttxy2))

print('\nTensões de corte maximas/mínima: -------------------------------------------')
aS1 = angS(sxx,syy,txy)/2 # ângulo no espaço físico, radianos
print('Ângulo tensão corte max (Deg) =','{:.2f}'.format(np.rad2deg(aS1)))
tcortexxMa = sxx1(sxx,syy,txy,aS1)
tcorteyyMa = syy1(sxx,syy,txy,aS1)
tcortexyMa = txy1(sxx,syy,txy,aS1)
print('tens xx - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcortexxMa))
print('tens yy - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcorteyyMa))
print('tens xy - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcortexyMa))

aS2 = aS1 + np.pi/2
print('Ângulo tensão corte min (Deg) =','{:.2f}'.format(np.rad2deg(aS2)))
tcortexxMb = sxx1(sxx,syy,txy,aS2)
tcorteyyMb = syy1(sxx,syy,txy,aS2)
tcortexyMb = txy1(sxx,syy,txy,aS2)
print('tens xx - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcortexxMb))
print('tens yy - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcorteyyMb))
print('tens xy - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcortexyMb))

# use LaTeX fonts in the plot
# plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams['axes.facecolor']='gray'
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(angt, tsxx, 'k--', label=r'$\sigma_{xx}$', linewidth=3)
ax.plot(angt, tsyy, 'k:', label=r'$\sigma_{yy}$', linewidth=3)
ax.plot(angt, ttxy, 'k', label=r'$\tau_{xy}$', linewidth=3)
# points: principal stresses
ax.plot(aprinc, tsxx1, 's'), ax.text(aprinc, tsxx1,r'$\sigma_{xx}$')
ax.plot(aprinc, tsyy1, 's'), ax.text(aprinc, tsyy1,r'$\sigma_{yy}$')
ax.plot(aprinc, ttxy1, 's'), ax.text(aprinc, ttxy1,r'$\tau_{xy}$')
ax.plot(aprinc2, tsxx2, 's'), ax.text(aprinc2, tsxx2,r'$\sigma_{xx}$')
ax.plot(aprinc2, tsyy2, 's'), ax.text(aprinc2, tsyy2,r'$\sigma_{yy}$')
ax.plot(aprinc2, ttxy2, 's'), ax.text(aprinc2, ttxy2,r'$\tau_{xy}$')
# points: critical shear stresses
ax.plot(np.rad2deg(aS1), tcortexxMa, 'v')
ax.text(np.rad2deg(aS1), tcortexxMa,r'$\varepsilon_{xx}$')
ax.plot(np.rad2deg(aS1), tcorteyyMa, 'v')
ax.text(np.rad2deg(aS1), tcorteyyMa,r'$\varepsilon_{yy}$')
ax.plot(np.rad2deg(aS1), tcortexyMa, 'v')
ax.text(np.rad2deg(aS1), tcortexyMa,r'$\gamma_{xy}$')
ax.plot(np.rad2deg(aS2), tcortexxMb, 'v')
ax.text(np.rad2deg(aS2), tcortexxMb,r'$\varepsilon_{xx}$')
ax.plot(np.rad2deg(aS2), tcorteyyMb, 'v')
ax.text(np.rad2deg(aS2), tcorteyyMb,r'$\varepsilon_{yy}$')
ax.plot(np.rad2deg(aS2), tcortexyMb, 'v')
ax.text(np.rad2deg(aS2), tcortexyMb,r'$\gamma_{xy}$')

ax.set(xlabel='Ângulo (deg)', ylabel='Transformação das tensões')
legend = ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., shadow=True, fontsize=18)
ax.axhline(y=0, color='k', linewidth=1)
ax.axvline(x=0, color='k', linewidth=1)
ax.grid()
plt.show()
```

### c) As direções principais.<a name="C"></a>
([enunciatum](#PI31))

As direções principais de tensão são obtidas analitacamente pela resolução do seguinte sistema de equações:

\begin{equation*}
(\sigma_{ij} - \sigma_\alpha\delta_{ij}) \cdot n_j = 0 \quad \wedge \quad n_i \cdot n_i = 1
\quad (\alpha = \mathrm{I,II,III})
\end{equation*}

```python
from sympy.solvers import solve
from sympy import symbols

nx,ny,nz = symbols('nx ny nz')

def eq1(l,m,n,vp): return (tt0[0,0] - vp)*l + tt0[0,1]*m + tt0[0,2]*n
def eq2(l,m,n,vp): return tt0[1,0]*l + (tt0[1,1] - vp)*m + tt0[1,2]*n
def eq3(l,m,n,vp): return tt0[2,0]*l + tt0[2,1]*m + (tt0[2,2] - vp)*n
def eq4(l,m,n): return l**2 + m**2 + n**2 - 1

def versor(e1,e2,e3): return np.array([e1,e2,e3])

# 1 valor proprio -------------------
print('-------------------------------------------')
print('Vetor próprio associado a sI = ','{:.2f}'.format(sI),' MPa')
print('-------------------------------------------')
print('\n(i) sistema de equações:')
E1_sI = eq1(nx,ny,nz,sI); print(E1_sI,' = 0')
E2_sI = eq2(nx,ny,nz,sI); print(E2_sI,' = 0')
E3_sI = eq3(nx,ny,nz,sI); print(E3_sI,' = 0')
E4_sI = eq4(nx,ny,nz); print(E4_sI,' = 0')
VetProp1 = solve([E1_sI,E2_sI,E4_sI], [nx,ny,nz], dict=True)
l1 = float(VetProp1[1][nx])
m1 = float(VetProp1[1][ny])
n1 = float(VetProp1[1][nz])
versorI = versor(l1,m1,n1)
print('\n(ii) nI = ','{:.3f}'.format(l1),'i +','{:.3f}'.format(m1),'j +','{:.3f}'.format(n1),'k')
```

```python
# 2 valor proprio -------------------
print('-------------------------------------------')
print('Vetor próprio associado a sII = ','{:.2f}'.format(sII),' MPa')
print('-------------------------------------------')
print('\n(i) sistema de equações:')
E1_sII = eq1(nx,ny,nz,sII); print(E1_sII,' = 0')
E2_sII = eq2(nx,ny,nz,sII); print(E2_sII,' = 0')
E3_sII = eq3(nx,ny,nz,sII); print(E3_sII,' = 0')
E4_sII = eq4(nx,ny,nz); print(E4_sII,' = 0')
VetProp2 = solve([E1_sII,E2_sII,E3_sII,E4_sII], [nx,ny,nz], dict=True)
l2 = float(VetProp2[1][nx])
m2 = float(VetProp2[1][ny])
n2 = float(VetProp2[1][nz])
versorII = versor(l2,m2,n2)
print('\n(ii) nII = (','{:.3f}'.format(l2),') i + (','{:.3f}'.format(m2),') j + (','{:.3f}'.format(n2),') k')
```

```python
# 3 valor proprio -------------------
print('-------------------------------------------')
print('Vetor próprio associado a sIII = ','{:.2f}'.format(sIII),' MPa')
print('-------------------------------------------')
print('\n(i) sistema de equações:')
E1_sIII = eq1(nx,ny,nz,sIII); print(E1_sIII,' = 0')
E2_sIII = eq2(nx,ny,nz,sIII); print(E2_sIII,' = 0')
E3_sIII = eq3(nx,ny,nz,sIII); print(E3_sIII,' = 0')
E4_sIII = eq4(nx,ny,nz); print(E4_sIII,' = 0')
VetProp3 = solve([E1_sIII,E2_sIII,E4_sIII], [nx,ny,nz], dict=True)
l3 = float(VetProp3[1][nx])
m3 = float(VetProp3[1][ny])
n3 = float(VetProp3[1][nz])
versorIII = versor(l3,m3,n3)
print('\n(ii) nIII = (','{:.3f}'.format(l3),') i + (','{:.3f}'.format(m3),') j + (','{:.3f}'.format(n3),') k')
```

De notar que a solução algébrica do sistema de equações linear homogéneo conduz a soluções do tipo:

\begin{equation*}
\vec{n_\mathrm{I}} = \pm l_1 \vec{\imath} \pm m_1 \vec{\jmath} \pm n_1 \vec{k}
\end{equation*}

Os versores das direções principais devem formar entre si um sistema de
coordenadas ortonormado e direto, ou seja, devem satisfazer as seguintes
relações:

- Sistema ortonormado:

\begin{equation*}
\begin{aligned}
e_\mathrm{I} \cdot e_\mathrm{I} =  1 &\quad\wedge\quad
e_\mathrm{II} \cdot e_\mathrm{II} =  1 &\wedge\quad
e_\mathrm{III} \cdot e_\mathrm{III} =  1 \\
e_\mathrm{I} \cdot e_\mathrm{II} =  0 &\quad\wedge\quad
e_\mathrm{I} \cdot e_\mathrm{III} =  0 &\wedge\quad~
e_\mathrm{II} \cdot e_\mathrm{III} =  0 \\
\end{aligned}
\end{equation*}

- Sistema direto (regra da mão direita da multiplicação de vetores)

\begin{equation*}
\begin{aligned}
e_\mathrm{I} \times e_\mathrm{II} &= e_\mathrm{III}\\
e_\mathrm{II} \times e_\mathrm{III} &= e_\mathrm{I} \\
e_\mathrm{III} \times e_\mathrm{I} &= e_\mathrm{II} \\
\end{aligned}
\end{equation*}

Simplificando, a definição de um sistema de coordenadas ortonormado e direto
deve satisfazer o seguinte produto misto entre versores dos eixos coordenados
principais:

\begin{equation*}
\vec{e}_\mathrm{I} \cdot ( \vec{e}_\mathrm{II} \times \vec{e}_\mathrm{III})  =
 \det([Q])
 = +1
\end{equation*}

#### c) As direções principais: pelo método da circunferência de Mohr.<a name="C"></a>
([enunciatum](#PI31))

No espaço das tensões da circunferência de Mohr ($\sigma,\tau$) o ângulo entre o referêncial
principal e o atual é dado por:

\begin{equation*}
\begin{aligned}
2\theta_{P1} &= \arctan \left
(\frac{2\tau_{yz}}{\sigma_{yy}-\sigma_{zz}}\right) \\
2\theta_{P2} &= 2\theta_{P1} + 2(90^\circ)\\
\end{aligned}
\end{equation*}

O ponto $I$ é obtido a partir do ponto $A$ por uma rotação $2\theta$ em torno do ponto $C$.

Atendendo a que:

\begin{equation*}
\vec{n}_\mathrm{II} \equiv \vec{\imath}_\mathrm{P}~;~
\vec{n}_\mathrm{III} \equiv \vec{\jmath}_\mathrm{P}~;~
\vec{n}_\mathrm{I} \equiv \vec{k}_\mathrm{P}
\end{equation*}

A matriz de transformação do sistema de coordenadas $S(O,\vec{\imath}\,
\vec{\jmath}\,\vec{k})$ (original) para o sitema de coordenadas
principal
$S_p(O, \vec{\imath}_\mathrm{P}\vec{\jmath}_\mathrm{P}\vec{k}_\mathrm{P})$
(novo) vem:

\begin{equation*}
\begin{array}{ccccc}
&& \vec{\imath} & \vec{\jmath} & \vec{k} \\
&& \downarrow & \downarrow & \downarrow \\
\vec{\imath}_\mathrm{P} & \rightarrow & \vec{\imath}_\mathrm{P} \cdot \vec{\imath}
& \vec{\imath}_\mathrm{P} \cdot \vec{\jmath} & \vec{\imath}_\mathrm{P} \cdot \vec{k} \\
\vec{\jmath}_\mathrm{P} & \rightarrow & \vec{\jmath}_\mathrm{P} \cdot \vec{\imath}
& \vec{\jmath}_\mathrm{P} \cdot \vec{\jmath} & \vec{\jmath}_\mathrm{P} \cdot \vec{k} \\
\vec{k}_\mathrm{P} & \rightarrow & \vec{k}_\mathrm{P} \cdot \vec{\imath} &
\vec{k}_\mathrm{P} \cdot \vec{\jmath} & \vec{k}_\mathrm{P} \cdot \vec{k}
\end{array}
\quad\quad\quad\quad
\begin{aligned}
\text{}\\[2.75ex]
\begin{cases}
\vec{\imath}_\mathrm{P} &= 1 \vec{\imath} + 0 \vec{\jmath} + 0 \vec{k} \\
\vec{\jmath}_\mathrm{P} &= 0 \vec{\imath} + \cos\theta_\mathrm{P} \vec{\jmath} + \sin\theta_\mathrm{P}\vec{k} \\
\vec{k}_\mathrm{P} &= 0 \vec{\imath} -\sin\theta_\mathrm{P} \vec{\jmath} + \cos\theta_\mathrm{P} \vec{k} \\
\end{cases}
\end{aligned}
\end{equation*}

em forma matricial,

\begin{equation*}
\begin{bmatrix}
\vec{\imath}_\mathrm{P} \\ \vec{\jmath}_\mathrm{P} \\ \vec{k}_\mathrm{P}
\end{bmatrix} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & \cos\theta_\mathrm{P} & \sin\theta_\mathrm{P} \\
0 & -\sin\theta_\mathrm{P} & \cos\theta_\mathrm{P} \\
\end{bmatrix}
\begin{bmatrix}
\vec{\imath}  \\ \vec{\jmath}  \\ \vec{k}
\end{bmatrix}
\end{equation*}

Se o ângulo $\theta$ for negativo, atendendo a que $T(-\theta)=T(\theta)^T$ a matriz anterior vem:

\begin{equation*}
\begin{bmatrix}
\vec{\imath}_\mathrm{P} \\ \vec{\jmath}_\mathrm{P} \\ \vec{k}_\mathrm{P}
\end{bmatrix} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & \cos\theta_\mathrm{P} & -\sin\theta_\mathrm{P} \\
0 & \sin\theta_\mathrm{P} & \cos\theta_\mathrm{P} \\
\end{bmatrix}
\begin{bmatrix}
\vec{\imath}  \\ \vec{\jmath}  \\ \vec{k}
\end{bmatrix}
\end{equation*}

```python
def angP(s11,s22,s12): return np.arctan((2*s12)/(s11-s22))

syy = tt0[1,1]
szz = tt0[2,2]
syz = tt0[1,2]

aP1 = angP(syy,szz,syz)
aP2 = aP1 + np.pi

print('angulos das direções principais')
print('\nNo espaço das tensões:')
print('angPI:  ','{:.3f}'.format(np.rad2deg(aP1)),' deg | ','{:.3f}'.format(aP1),' radians')
print('angPII: ','{:.3f}'.format(np.rad2deg(aP2)),' deg | ','{:.3f}'.format(aP2),' radians')

aPI = aP1/2
aPII = aP2/2

print('\nNo espaço fisico:')
print('angPI:  ','{:.3f}'.format(np.rad2deg(aPI)),' deg | ','{:.3f}'.format(aPI),' radians')
print('angPII: ','{:.3f}'.format(np.rad2deg(aPII)),' deg | ','{:.3f}'.format(aPII),' radians')

def Q(ang,rot):
    c = np.cos(ang)
    s = np.sin(ang)
    if rot == 1:
        Qij = np.array([[1,0,0],[0,c,s],[0,-s,c]])
    elif rot == 2:
        Qij = np.array([[c,0,s],[0,1,0],[-s,0,c]])
    else:
        Qij = np.array([[c,s,0],[-s,c,0],[0,0,1]])
    return Qij

MatrTrans = Q(aPI,1)
print('\nMatriz de transformação = \n',MatrTrans)

print('\nVetores principais:')
nI = MatrTrans[0,:]
print('eI = ', nI)
nII = MatrTrans[1,:]
print('eII = ', nII)
nIII = MatrTrans[2,:]
print('eIII = ', nIII)
```

Verificação da transformação:

\begin{equation*}
\begin{aligned}
\underline{\underline{\sigma}}_P&=\underline{\underline{T}}_{0P}\,\underline{\underline{\sigma}}_0~
\underline{\underline{T}}_{0P}^T
\end{aligned}
\end{equation*}

```python
np.set_printoptions(precision=3,suppress=True)

sprinc = np.dot(MatrTrans,np.dot( tt0 ,np.transpose(MatrTrans) ))
print('Matriz das tensões no refencial principal = \n',sprinc)
```

### d) A tensão tangencial máxima.<a name="D"></a>
([enunciatum](#PI31))

A análise do círculo de Mohr permite determinar os valores máximos e mínimos das componentes de
corte das tensões num determinado plano de análise. Considere-se esse plano $Oyz$, vem então:

\begin{equation*}
\begin{aligned}
\tau_\mathrm{max/min} & \equiv R = \pm \sqrt{\left(\frac{\sigma_{yy}-\sigma_{zz}}{2}\right)^2 +
\tau_{yz}^2 }
\end{aligned}
\end{equation*}

```python
syy = tt0[1,1]
szz = tt0[2,2]
syz = tt0[1,2]

Raio = np.sqrt(((syy-szz)/2)**2 + syz**2)

tau_max = Raio
print('tau_max = ','{:.2f}'.format(tau_max),' MPa')
tau_min = -Raio
print('tau_min = ','{:.2f}'.format(tau_min),' MPa')
```

### e) Os ângulos que o eixo perpendicular ao plano onde ocorre a tensão tangencial máxima faz com os eixos do referencial $xyz$.<a name="E"></a>
([enunciatum](#))

Os ângulos que o eixo perpendicular ao plano onde ocorre a tensão tangencial máxima faz
com os eixos do referencial $xyz$ é dado por:

\begin{equation*}
\begin{aligned}
\theta_{S}^\textrm{Morh} = 2\theta_{S}^\textrm{Fisico} &= \arctan \left
(-\frac{\sigma_{yy}-\sigma_{zz}}{2\tau_{yz}}\right) \\[1ex]
\end{aligned}
\end{equation*}

```python
def angS(s11,s22,s12): return -np.arctan((s11-s22)/(2*s12))

aS1 = angS(syy,szz,syz)
aS2 = aS1 + np.pi

print('angulos das direções principais')
print('\nNo espaço das tensões:')
print('angSI:  ','{:.3f}'.format(np.rad2deg(aS1)),' deg | ','{:.3f}'.format(aS1),' radians')
print('angSII: ','{:.3f}'.format(np.rad2deg(aS2)),' deg | ','{:.3f}'.format(aS2),' radians')

aSI = aS1/2
aSII = aS2/2

print('\nNo espaço fisico:')
print('angPI:  ','{:.3f}'.format(np.rad2deg(aSI)),' deg | ','{:.3f}'.format(aSI),' radians')
print('angPII: ','{:.3f}'.format(np.rad2deg(aSII)),' deg | ','{:.3f}'.format(aSII),' radians')
```
