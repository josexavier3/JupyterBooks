## Problema 1 <a name="PIV91"></a>

A viga rígida $ABCD$, representada na figura abaixo, está simplesmente apoiada em $B$ e está ligada por articulações a 3 barras idênticas, $AE$, $CF$ e $DG$. Não existem tensões iniciais. A barra $AE$ é submetida a um aumento de temperatura de 20ºC para 220ºC. Determine o deslocamento da extremidade $D$ da viga e as tensões nas restantes barras.

Dados: $E_\mathrm{barra} = $ 210 GPa, $\alpha_\mathrm{barra} = 12\times10^{-6}$ ºC $^{-1}$, $S_\mathrm{barra} = 300$ mm $^2$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au09/Au09_P001_i1.jpg"
width="550">


```python
# unidades: mm, N, MPa
# dados
E = 210.e3 # unidade: MPa
A = 300. # unidade: mm2
alphaAE = 12.e-6 # unidade: º/C
LAE = 300. # unidade: mm
LCF = LAE
LDG = LAE
LAB = 250.
LBC = LAB
LCD = LAB
LBD = 2*LAB
Ti, Tf = 20., 220. # unidades: ºC
DT = Tf - Ti
print(f'DT = {DT:.1f} ºC')
```

### Resolução

**Problemas estaticamente indeterminados**

Existem problemas práticos de engenharia, para os quais as incógnitas de forças internas e reações não podem ser determinadas apenas pela análise do equilíbrio estático (Mecânica Aplicada I. Estática). Ou seja, a resolução do problema não conduz a um sistema de equações resolúvel e único, obtido a partir da análise de forças externas representadas no diagrama de corpo livre (DCL) do sistema mecânico.

Neste caso, as equações de equilíbrio podem ser complementadas por relações cinemáticas (deformações) considerando a geometria do problema.

**Análise do problema**

A estrutura do enunciado é estaticamente indeterminada, pois existem mais incógnitas (esforços nas barras e reações) do que equações de equilíbrio estático nos nós da estrutura (num problema plano: duas equações da resultante das forças e uma equação da resultante dos momentos). A solução do problema recorre ao princípio da sobreposição, válido para estruturas ou mecanismos do domínio linear elástico.


A hipótese de que a barra $ABCD$ é rígida (i.e., deformação desprezável) permite confinar o seu movimento a uma rotação de corpo rígido, em torno do ponto $B$. Por sua vez, as barras verticais $AE$, $CF$ e $DG$ irão estar submetidas apenas a deslocamentos e esforços axiais (de tração ou de compressão).

Considere a aplicação do princípio da sobreposição de acordo com o esquema da figura. Neste caso, foi removida uma ligação ao exterior passando-se de uma estrutura hiperestática para uma estrutura isostática equivalente (este é o princípio de base do método das forças em mecânica das estruturas). Para garantir compatibilidade de deslocamentos a estrutura isostática irá ser solicitada pelo carregamento real mais uma força redundante aplicada no ponto e com a direção da reação do encastramento removido (de notar que a força redundante é a força de reação, na interpretação física do problema).

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au09/Au09_P001_i2.jpg"
width="600" height="350">

Neste problema foi removido o encastramento na secção $E$ (de notar que no contexto das hipóteses assumidas as forças de reação no encastramento equivalem apenas a 1 força axial ao longo da barra $AE$). Esta estrutura dita isostática equivalente para o problema em análise ficará submetida à solicitação real (variação de temperatura da barra $AE$) mais um carregamento onde é aplicada uma força dita redundante (que fisicamente representa o efeito do encastramento em $E$).

Quando ocorre variação (aumento) de temperatura da barra $AE$, esta irá sofrer um alongamento $\delta_T$ dado por ($\Delta T >0~\Rightarrow \varepsilon >0,~\delta >0$):

\begin{equation*}
\delta_{E}^T \equiv \delta_T = \alpha \Delta T L
\end{equation*}

De notar que a barra rígida $ABCD$, bem como as duas barras $EF$ e $DG$, não sofrem qualquer deformação, neste caso. Este cenário é equivalente a assumir um encastramento na secção $A$, pelo que o deslocamento $\delta_T$ fica representado no diagrama como um deslocamento vertical no ponto $E$, ou, por outras palavras, um deslocamento axial da barra na secção $E$ de origem térmica.

```python
deltaT = alphaAE*DT*LAE
print(f'deltaT = {deltaT:.3f} mm')
```

Por uma questão de compatibilidade de deslocamentos (de notar que na estrutura real, o deslocamento no ponto $E$ é nulo devido ao encastramento) a secção $E$ deve ser submetida a uma força $F_{AE}$ (força redundante) que permite alunar o deslocamento gerado pela variação de temperatura (por efeito térmico). De notar que esta forma é compressiva! Essa força irá, em consequência, deformar mecanicamente a restante estrutura, como se tratando de uma força externa aplicada (ver esquema da figura). De notar que a força axial de compressão na barra $AE$ também irá gerar forças de compressão nas barras $CD$ e $DB$.

Isolando a barra $AE$ têm-se a sobreposição dos efeitos de origem térmicos e mecânicos. Pelo que o deslocamento efetivo da barra será dado por:

\begin{equation*}
\delta_{E/A}=\delta_E^F-\delta_A^F = \left(\frac{F L}{AE}\right)_{AE}
\quad\wedge\quad
\delta_E^F \equiv \delta_E^T~(\textrm{encastramento})
\end{equation*}

**Equilíbrio**

Considere-se  seguinte diagrama de corpo livre da barra rigida:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au09/Au09_P001_i3.jpg"
width="380" height="175">

Desenvolvendo as expressões de equilíbrio estático vem:

\begin{equation*}
\begin{cases}
+/\uparrow~&:~\sum F_y=0
\quad\Leftrightarrow\quad
R_\mathrm{By} = F_{AE} + F_{CF} + F_{DG} \\
+/\curvearrowleft~&:~\sum M_B=0
\quad\Leftrightarrow\quad
L_\mathrm{AB} F_\mathrm{AE} - L_\mathrm{BC} F_\mathrm{CF} - L_\mathrm{BD} F_\mathrm{DG} =0 \\
\end{cases}
\end{equation*}

Da expressão do equilíbrio dos momentos das forças aplicadas em relação ao ponto $B$ resulta ($L_{AB} = L_{BC}$, $L_{BD} = 2L_{AC}$):

\begin{equation*}
\sum M_B=0 \quad\Leftrightarrow\quad
F_\mathrm{AE} = F_\mathrm{CF} + 2F_\mathrm{DG}\\
\end{equation*}

De notar que será possível demostrar, nomeadamente pela compatibilidade de deslocamentos que se verifica a sesuinte relação: $F_\mathrm{DG} = 2F_\mathrm{CF}~\therefore~F_\mathrm{AE}=5 F_\mathrm{CF}$

**Compatibilidade de deslocamentos**

Da análise da geometria da deformada da estrutura (válida pela hipótese da barra horizontal rígida), resulta pela análise da semelhança entre triângulos (ver figura):

\begin{equation*}
\frac{\delta_A}{L_{AB}}  = \frac{\delta_C}{L_{BC}} = \frac{\delta_D}{L_{BD}}~,
\quad\textrm{em que } \delta\equiv|\delta|
\end{equation*}

de onde se conclui ($L_{AB} = L_{BC}$, $L_{BD} = 2L_{AC}$ ):

\begin{equation*}
\delta_A = \delta_C
\quad\wedge\quad
\delta_C = \frac{\delta_D}{2}
\end{equation*}

**Relações força-deslocamento**

Os deslocamentos das secções $C$ e $D$ são apenas de origem elástica ou seja:

\begin{equation*}
\delta_C = \frac{F_{CF} L}{AE} \quad\wedge\quad \delta_D = \frac{F_{DG} L}{AE}
\end{equation*}

Da combinação das duas últimas equações resulta a seguinte relação entre esforços axiais:

\begin{equation*}
\delta_C = \frac{\delta_D}{2}
\quad\Rightarrow \quad
F_{CF} = \frac{F_{DG}}{2}
\end{equation*}

Se a estrutura está em equilíbrio estático, isso significa que também os elementos que a constituem (nós e barras) estão em equilíbrio. Para este problema deve-se considerar o diagrama de corpo livre de cada elemento (barra), indicando sobre este todas as forças que nele atuam. Considere-se, nesta etapa, o diagrama de corpo livre da barra rígida $ABCD$. Pelas forças de reação nas secções $E$, $F$ e $G$ as barras $AE$, $CF$ e $DG$ estão submetidas a esforços axiais de compressão. Pelo que no diagrama de corpo livre da barra rigida estas forças aparecem com sentido vertical positivo (contrariando o efeito da deformação das barras). Neste caso, represente-se a reação vertical na secção $B$ como negativa.

**Solução final**

Da equação de deslocamento relativo (alongamento efetivo) da barra $AE$ vem:

\begin{equation*}
\begin{aligned}
\delta_E - \delta_A  &= \frac{F_{AE} L }{AE} \quad \longleftarrow \quad \delta_A = \delta_C;~ F_\mathrm{AE} = 5F_{CF} \\
\alpha \Delta T L - \delta_C  &= \frac{5F_{CF} L }{AE} \\
\alpha \Delta T L - \frac{F_{CF} L }{AE}  &= \frac{5F_{CF} L }{AE} \\
F_{CF} &= \frac{\alpha \Delta T AE}{6}
\end{aligned}
\end{equation*}

resulta ainda que,

\begin{equation*}
\begin{aligned}
F_{AE} &= F_{CF} + 2 (2F_{CF}) = 5F_{CF} \\
F_{DG} &= 2F_{CF}\\
\end{aligned}
\end{equation*}

```python
FCF = alphaAE*DT*A*E/6
print(f'FCF = {FCF:.1f} N')

FAE = 5*FCF
print(f'FAE = {FAE:.1f} N')

FDG = 2*FCF
print(f'FDG = {FDG:.1f} N')

sigCF = FCF/A
print(f'sigCF = {sigCF:.1f} MPa')

sigAE = FAE/A
print(f'sigAE = {sigAE:.1f} MPa')

sigDG = FDG/A
print(f'sigDG = {sigDG:.1f} MPa')

deltaAE = FAE*LAE/A/E
print(f'deltaAE = {deltaAE:.3f} mm')

deltaA = LAE*(alphaAE*DT - FAE/A/E)
print(f'deltaA = {deltaA:.3f} mm')

deltaC = deltaA
print(f'deltaC = {deltaC:.3f} mm')

deltaD = 2*deltaC
print(f'deltaD = {deltaD:.3f} mm')
```
