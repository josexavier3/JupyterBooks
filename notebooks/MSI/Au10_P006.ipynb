## Tração e compressão de barras elásticas e elastoplásticas perfeitas
## Vigas sujeitas à tracção/compressão. Parte III
## Problema 6 <a name="PIV106"></a>


A figura representa um conjunto constituído por um parafuso de aço com 8 mm de diâmetro, um tubo de latão com diâmetros exterior e interior respectivamente iguais a $D_e$= 14 mm e $D_i$= 12 mm e uma porca de aço. O parafuso foi colocado dentro do tubo e a porca foi perfeitamente ajustada de forma a não haver folgas entre a cabeça do parafuso e o tubo. Sabe-se que os dois materiais têm comportamento elastoplástico perfeito, com as seguintes propriedades, $E(\mathrm{aço})$= 210 GPa, $E(\mathrm{latão})$= 105 GPa, $\sigma_C(\mathrm{aço})$ = 235 MPa, $\sigma_C(\mathrm{latão})$ = 320 MPa, que $L$ = 200 mm e que o passo do parafuso é de 0,775 mm. Determine a rotação máxima que é possível aplicar à porca para que não haja cedência plástica de qualquer um dos materiais.
