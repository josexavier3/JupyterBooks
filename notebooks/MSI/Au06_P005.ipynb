## Resolução

#### a) A variação da altura e do volume para o carregamento indicado $\sigma_{yy} = -83$ MPa.<a name="A"></a>

([enunciatum](#PIII65))

##### Da lei de Hooke para um material isotrópico.

Para um material isotrópico a lei de Hooke tridimensional vem expressa por:

\begin{equation*}
\varepsilon_{ij} = \frac{1}{E}[(1+\nu)\sigma_{ij} - \nu\delta_{ij}\sigma_{kk}]
\end{equation*}

Em forma matricial a lei constitutiva linear elástica expressa em termos de módulo de Young e coeficiente de Poisson
vem:

\begin{equation*}
\begin{aligned}
\varepsilon_{11} &= \displaystyle\frac{1}{E}[\sigma_{11} - \nu(\sigma_{22}+\sigma_{33})] \\
\varepsilon_{22} &= \displaystyle\frac{1}{E}[\sigma_{22} - \nu(\sigma_{11}+\sigma_{33})] \\
\varepsilon_{33} &= \displaystyle\frac{1}{E}[\sigma_{33} - \nu(\sigma_{11}+\sigma_{22})] \\
2\varepsilon_{23} &= \gamma_{23} = \displaystyle\frac{\tau_{23}}{G}\\
2\varepsilon_{13} &= \gamma_{13} = \displaystyle\frac{\tau_{13}}{G}\\
2\varepsilon_{12} &= \gamma_{12} = \displaystyle\frac{\tau_{12}}{G}\\
\end{aligned}
\end{equation*}

com,

\begin{equation*}
G = \frac{E}{2(1+\nu)}
\end{equation*}

Na forma matricial têm-se:

\begin{equation*}
\begin{bmatrix}
\varepsilon_{11} \\ \varepsilon_{22} \\ \varepsilon_{33} \\ 2\varepsilon_{23} \\ 2\varepsilon_{13} \\ 2\varepsilon_{12}
\end{bmatrix} = \frac{1}{E}
\begin{bmatrix}
1 & -\nu & -\nu & 0 & 0 & 0 \\
-\nu & 1 & -\nu & 0 & 0 & 0 \\
-\nu & -\nu & 1 & 0 & 0 & 0 \\
0 & 0 & 0 & 2(1+\nu) & 0 & 0 \\
0 & 0 & 0 & 0 & 2(1+\nu) & 0 \\
0 & 0 & 0 & 0 & 0 & 2(1+\nu) \\
\end{bmatrix}
\begin{bmatrix}
\sigma_{11} \\ \sigma_{22} \\ \sigma_{33} \\ \tau_{23} \\  \tau_{13} \\ \tau_{12}
\end{bmatrix}
\end{equation*}

A segunda lei de termodinâmica requer que as constantes elásticas verifiquem os seguintes limites:

\begin{equation*}
\begin{aligned}
E,~G > 0 \\
-1 < \nu < \frac{1}{2}
\end{aligned}
\end{equation*}

Para o problema proposto, o estado de tensão é dado por:

\begin{equation*}
\underline{\underline{\sigma}} = \sigma_{ij} =
\begin{bmatrix}
0 & 0 & 0 \\
0 & \sigma_{yy} & 0 \\
0 & 0 & 0 \\
\end{bmatrix} =
\begin{bmatrix}
0 & 0 & 0 \\
0 & -83 & 0 \\
0 & 0 & 0 \\
\end{bmatrix}~\textrm{(MPa)}
\end{equation*}

As correspondentes componentes do tensor das deformações vêm:

\begin{equation*}
\begin{aligned}
\varepsilon_{xx} &= -\displaystyle\frac{\nu}{E}\sigma_{yy} \\
\varepsilon_{yy} &= \displaystyle\frac{1}{E}\sigma_{yy}\\
\varepsilon_{zz} &= -\displaystyle\frac{\nu}{E}\sigma_{yy} \\
\gamma_{yz} & = \gamma_{xz} = \gamma_{xy} = 0\\
\end{aligned}
\end{equation*}

- A variação de altura vem:

Por definição de deformação linear:

\begin{equation*}
\varepsilon = \frac{\Delta L}{L} = \frac{L_f - L_0}{L_0}
\end{equation*}

A deformação linear na direção do carregamento (direção ativa), o eixo dos $yy$, vem:

\begin{equation*}
\Delta L = \varepsilon_{yy} L
\end{equation*}

- A variação de volume vem:

Por definição a deformação volumétrica é dada por:

\begin{equation*}
\begin{aligned}
\varepsilon_\mathrm{V} &= \frac{\Delta V}{V} =  \mathrm{tr}(\varepsilon)
= \varepsilon_{xx} + \varepsilon_{yy} + \varepsilon_{zz} \\
&= \frac{1-2\nu}{E}(\sigma_{xx} + \sigma_{yy} + \sigma_{zz})
= \frac{1-2\nu}{E}\sigma_{yy} = \frac{1-2\nu}{E}P
\end{aligned}
\end{equation*}

Sendo que o volume inicial não deformado $V$:

\begin{equation*}
V = \pi r^2 h
\end{equation*}

```python
import numpy as np

np.set_printoptions(precision=10,suppress=True)
E = 200*1e3  # unit: MPa
v = 0.3
G = E/2/(1+v)
print('Propriedades do material: ---------------')
print('     E (MPa) = ',E)
print('           v = ',v)
print(f'     G (MPa) = {G:.2f}')

appliedStress = -83  # unit: MPa

def StressTensor(s11,s12,s13,s22,s23,s33):
    return np.array([[s11,s12,s13],[s12,s22,s23],[s13,s23,s33]])

def StressVoigt(s11,s22,s33,s23,s13,s12):
    return np.array([s11,s22,s33,s23,s13,s12])

def C3Diso(E,v):
    return 1/E*np.array([[1,-v,-v,0,0,0],[-v,1,-v,0,0,0],[-v,-v,1,0,0,0],[0,0,0,2*(1+v),0,0],[0,0,0,0,2*(1+v),0],[0,0,0,0,0,2*(1+v)]])

stress = StressVoigt(0,appliedStress,0,0,0,0)
strain = np.dot(C3Diso(E,v),stress)
print('stress - Voigt = \n',stress)
print('(C3Diso(E,v) = \n',(C3Diso(E,v)))
print('strain = \n',strain)
exx = float(strain[0])
eyy = float(strain[1])
ezz = float(strain[2])
print(f'exx = {exx:.3e}')
print(f'eyy = {eyy:.3e}')
print(f'ezz = {ezz:.3e}')

# dimensões: unit: mm
d, h = 100, 160 
r = d/2
print('---------------------------------------------------.-')
print('Variação de comprimento para o carregamento indicado:')
print('---------------------------------------------------.-')
DLa = eyy*h
print(f'DLa = {DLa:.5f} mm')
```

```python

print('\n------------------------------------------------')
print('Variação de volume para o carregamento indicado:')
print('------------------------------------------------')

V0 = np.pi*r**2*h
eVa = (1 - 2*v)/E*appliedStress
print(f'\neVa = (1-2v)\\E*Syy= {eVa:.5e} = {eVa*1e6:.1f} micro-def')
DVa = eVa*V0
print('\nDelta V = ','{:.1f}'.format(DVa),' mm^3')
```

#### b) Resolva a alínea anterior mas supondo um carregamento hidrostático $\sigma_{xx}=\sigma_{yy}=\sigma_{zz} = -83$ MPa.<a name="B"></a>

([enunciatum](#PIII65))

A condição de carregamento hidrostático: $\sigma_{xx} = \sigma_{yy} = \sigma_{zz} \equiv \sigma = -83$ MPa, correspondente à pressão hidrostática.

\begin{equation*}
\begin{aligned}
\varepsilon_{xx} &= \varepsilon_{yy} = \varepsilon_{zz} =
\sigma\left(\frac{1}{E} - \frac{\nu}{E} - \frac{\nu}{E}\right) = \frac{1-2\nu}{E}P\\
\gamma_{yz} & = \gamma_{xz} = \gamma_{xy} = 0\\
\end{aligned}
\end{equation*}

- A variação de altura vem:

\begin{equation*}
\Delta L = \varepsilon_{yy} L = \left( \frac{1-2\nu}{E}P \right) L
\end{equation*}

- A variação de volume vem:

Por definição a deformação volumétrica é dada por:

\begin{equation*}
\varepsilon_\mathrm{V} = \frac{\Delta V}{V} =  \mathrm{tr}(\varepsilon)
= \varepsilon_{xx} + \varepsilon_{yy} + \varepsilon_{zz}
= 3 \left( \frac{1-2\nu}{E} \right) P = \frac{P}{K}
\end{equation*}

em que $K$ é o módulo de Bulk ou de compressibilidade volumétrica,

\begin{equation*}
K = \frac{E}{3(1-2\nu)}
\end{equation*}

pelo que,

\begin{equation*}
\Delta V  =  \varepsilon_\mathrm{V} V = \frac{P}{K}V
\end{equation*}

```python
print('---------------------------------------------------.-')
print('Variação de comprimento para o carregamento indicado:')
print('---------------------------------------------------.-')
DLb = (1-2*v)/E*appliedStress*h
print('Delta L = ','{:.4f}'.format(DLb),' mm')
```

```python

print('\n------------------------------------------------')
print('Variação de volume para o carregamento indicado:')
print('------------------------------------------------')
K = E/3/(1-2*v)
print('K = ','{:.2f}'.format(K), 'MPa')
eVb = appliedStress/K
print('\neVb = P\\K = ','{:.5e}'.format(eVb))
print('          = ','{:.1f}'.format(eVb*1e6),' micro-deformações')
DVb = eVb*V0
print('\nDelta V = ','{:.1f}'.format(DVb),' mm^3')
```
