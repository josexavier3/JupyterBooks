O tubo de 2 m de comprimento representado na figura é feito de um material elastoplástico que apresenta o diagrama tensão tangencial-distorção indicado na figura.

[a)](#A) Determine o momento de torção $T$ que provoca na superfície exterior do tubo a distorção $\gamma$ = 0,006 rad.

[b)](#B) Obtenha o ângulo de torção permanente do tubo quando esse momento $T$ for removido.

[c)](#C) Represente a distribuição de tensões tangenciais residuais no tubo.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P006_i1.jpg"
width="600">

```python
import numpy as np

# Dados
# unidades: m, Pa, N

L, ri, re = 2., 30.*1e-3, 35.*1e-3 # m
TauC = 210.e6 # Pa
gammaC, gamma = 0.003, 0.006
```

#### a) Determine o momento de torção $T$ que provoca na superfície exterior do tubo a distorção $\gamma$ = 0,006 rad.<a name="A"></a> ([_enunciatum_](#PV126))

A partir do diagrama tensão de corte - distorção angular, $\tau-\gamma$, definido para o material elasto-plástico, é possível determinar o diagrama momento torsor - ângulo de torção, $T-\phi$.

Para a tensão de cedência, $\tau_C = 210$ MPa, corresponderá um momento torsor elástico máximo assim definido (para o veio oco: $c=r_e$, pois é no ponto da periferia que inicialmente ocorre cedência):

\begin{equation*}
T_C = \frac{\tau_C J}{r_e}
\end{equation*}

em que $J$ é o momento polar de inércia de secção; para um veio oco é expresso por:

\begin{equation*}
J = \frac{\pi}{2}(r_e^4 - r_i^4)
\end{equation*}

```python
J = np.pi/2*(re**4 - ri**4)
print(f'J = {J:.3e} m3')
Tc = TauC*J/re
print(f'Tc = {Tc:.2f} N.m')
```

O módulo de corte do material de que é feito o veio representa o declive, no troço linear elástico, na lei constitutiva $\tau-\gamma$:

\begin{equation*}
G = \frac{\tau_C}{\gamma_C}
\end{equation*}

```python
G = TauC/gammaC
print(f'G = {G*1e-9:.1f} GPa')
```

Associado ao momento elástico máximo, $T_C$, existirá o ângulo de torção crítico (limite de elasticidade), $\phi_C$. Este pode ser determinado de acordo com a expressão:

\begin{equation*}
\phi_C = \frac{T_C L}{JG}
\end{equation*}

```python
phiC = Tc*L/J/G
print(f'phiC = {phiC:.4f} radianos | {np.rad2deg(phiC):.2f} º')
```

Para a distorção angular $\gamma$ = 0.006 radianos, é possível determinar o ângulo de torção correspondente de acordo com as seguintes relações:

\begin{equation*}
\gamma L = \phi \rho
~\Leftrightarrow~
\phi = \frac{\gamma L}{r_e}
\end{equation*}

```python
phi = gamma*L/re
print(f'phi = {phi:.4f} radianos | {np.rad2deg(phi):.2f} º')
```

É possível determinar o raio que define o núcleo de material no domínio elástico através da relação:

\begin{equation*}
\frac{\phi_C}{\rho_C}  = \frac{\phi}{\rho}
~\Leftrightarrow~
\rho_C = \frac{\phi_C}{\phi} r_e
\end{equation*}

```python
rhoC = gammaC/gamma*re
print(f'rhoC = {rhoC*1e3:.2f} mm')
```

O valor calculado para $\rho_C$ permite concluir que o vei oco está completamente plastificado: $\rho_C < r_i$. O momento de torção para o domínio plástico para um veio oco é expresso por:

\begin{equation*}
\begin{aligned}
T_P & = \int_A\rho\,\tau\,\mathrm{d}A = \int_{r_i}^{r_e}\int_0^{2\pi}\rho\,\tau\,(\rho\mathrm{d}\theta\mathrm{d}\rho) \\
& = \int_{r_i}^{r_e}\rho^2\,\tau[\theta]_0^{\,2\pi} \mathrm{d}\rho
= \left[\frac{\rho^3}{3}\right]_{r_i}^{r_e}\,\tau\,2\pi \mathrm{d}\rho
= \frac{2\pi}{3}(r_e^3-r_i^3)\tau\\
\end{aligned}
\end{equation*}

Pelo que o momento de torção $T$ que provoca na superfície exterior do tubo a distorção $\gamma$ = 0,006 rad vem dado por:

\begin{equation*}
T \equiv T_P = \frac{2\pi}{3}\tau_C(r_e^3-r_i^3)
\end{equation*}

```python
Tp = 2*np.pi/3*TauC*(re**3-ri**3)
print(f'Tp = {Tp:.2f} N.m')
```

#### b) Obtenha o ângulo de torção permanente do tubo quando esse momento $T$ for removido.<a name="B"></a> ([_enunciatum_](#PV126))

No diagrama momento torsor - ângulo de torção ($T-\phi$) a parte de descarga do momento aplicado será representado por um segmento de reta, com declive igual ao troço linear. De notar que o ângulo de torção não volta a zero depois da remoção do momento torsor aplicado, resultando uma deformação angular permanente obtida por:

\begin{equation*}
\begin{aligned}
\phi' = \phi - \phi_R = \frac{T_p L }{JG}
\end{aligned}
\end{equation*}

em que,

- $\phi$ é o ângulo de torção para o momento torsor máximo aplicado
- $\phi'$ é o ângulo de torção da recuperação elástica do veio na fase de descarga ($T_pL/JG$).

```python
phidescarga  = Tp*L/J/G
print(f'phi(descarga) = {phidescarga:.4f} radianos | {np.rad2deg(phidescarga):.2f} º')

phiresidual = phi - phidescarga
print(f'phiresidual = {phiresidual:.4f} radianos | {np.rad2deg(phiresidual):.2f} º')
```

#### c) Represente a distribuição de tensões tangenciais residuais no tubo.<a name="C"></a> ([_enunciatum_](#PV126))

As tensões residuais no material elastoplástico são obtidas por aplicando o princípio de superposição.

Por um lado, consideram-se as tensões devidas à aplicação do momento torsor, $T$. Por outro lado, são consideradas as tensões devido ao momento de torção, igual e oposto, que é aplicado na fase de descarga.

O primeiro grupo de tensões reflete o comportamento elasto-plástico do material durante a fase de carregamento. O segundo grupo tem o comportamento linear do mesmo material durante o estágio de descarga. A adição dos dois grupos de tensões fornece a distribuição das tensões residuais no eixo.

Na descarga, a tensão mínima ($\tau_\mathrm{min}$) e máxima ($\tau_\mathrm{max}$) lineares no interior ($r_i$) e exterior ($r_e$) do tubo vêm:

\begin{equation*}
\begin{cases}
\tau_\mathrm{max} &= \displaystyle\frac{(-T_P) r_e}{J} ~,&\quad \rho = r_i\\
\tau_\mathrm{min} &= \displaystyle\frac{(-T_P) r_i}{J} =
\tau_\mathrm{max} \displaystyle\frac{r_i}{r_e} ~,&\quad \rho = r_e\\
\end{cases}
\end{equation*}

```python
taumaxDescarga =  -Tp*re/J
print(f'tau max (Descarga) = {taumaxDescarga*1e-6:.2f} MPa')

tauminDescarga =  -Tp*ri/J
print(f'tau min (Descarga) = {tauminDescarga*1e-6:.2f} MPa')

tauminDescarga2 =  taumaxDescarga*ri/re
print(f'tau min (Descarga) (versão 2) = {tauminDescarga2*1e-6:.2f} MPa')
```

Assim sendo as tensões residuais têm uma distribuição linear cujos valores no ponto interior e exterior ao veio oco são
determinados por:

\begin{equation*}
\begin{cases}
\tau_i^R = \tau_C + \tau_\mathrm{min} \\
\tau_e^R = \tau_C + \tau_\mathrm{max}
\end{cases}
\end{equation*}

```python
tauresidualri = TauC + tauminDescarga
print(f'tau Residual (ri) = {tauresidualri*1e-6:.2f} MPa')

tauresidualre = TauC + taumaxDescarga
print(f'tau Residual (re) = {tauresidualre*1e-6:.2f} MPa')
```

De notar que as tensões residuais são positivas ou negativas em zonas distintas na secção do veio. Esta condição é necessária para se verificar a condição de que:

\begin{equation*}
T = \int \rho(\tau\mathrm{d}A) = 0
\end{equation*}
