## Resolução

#### a) Determine o momento torsor máximo $T$ que é possível aplicar em $A$, $T_\mathrm{max} \,=\, ?$. ([_enunciatum_](#PV113))

**Fórmula de torção**

A fórmula de torção estabelece uma relação de proporcionalidade entre o momento torsor aplicado a um veio, $T$, e a tensão de corte, $\tau$, de acordo com a seguinte relação:

\begin{equation*}
\tau = \left(\frac{T}{J}\right)\rho\quad\textrm{(Pa)}
\end{equation*}

em que,

- $T$ é o momento torsor (N.m)
- $\rho$ é a coordenada radial da posição do veio em relação ao seu centro (m)
- $J$ é o momento polar de 2ª ordem de área de secção tranversal (m $^4$). Para uma secção massiça circular:

    \begin{equation*}
    J^\bigcirc = \frac{\pi c^4}{2} \quad (\textrm{m}^4)
    \end{equation*}

A tensão de corte máxima ocorre na periferia do veio:

\begin{equation*}
\tau_\mathrm{max} = \frac{T c}{J} = \frac{2T}{\pi c^3}
\quad\textrm{(Pa)}
\end{equation*}

A inequeção de dimensionamento à torção escreve-se neste caso:

\begin{equation*}
\tau_\mathrm{max} \le \tau_\mathrm{adm}
~\Leftrightarrow~
\frac{T c}{J} \le \tau_\mathrm{adm}
~\Leftrightarrow~
T_\mathrm{max} \le \frac{\pi c^3 \tau_\mathrm{adm}}{2}
\quad\textrm{(N.m)}
\end{equation*}

```python
import numpy as np

# unidades: mm, MPa, N

LAB, LDE = 900., 500. # unidade: mm
rB, rD = 60., 90. # unidade: mm

d = 30.
r = d/2

Sadm, G = 163., 77.e3 # unidade: MPa

J = np.pi/2*r**4
print(f'J = {J:.2f} mm^4')

Tmax = Sadm*J/r
print(f'Tmax = {Tmax:.2f} N.mm | {Tmax*1e-3:.2f} N.m')

# ou
Tmax2 = np.pi*r**3*Sadm/2
print(f'Tmax (ver 2) = {Tmax2:.2f} N.mm | {Tmax2*1e-3:.2f} N.m')


import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 18})
# use LaTeX fonts in the plot
# plt.rc('text', usetex=True)
plt.rc('font', family='serif')
x, y = r, Sadm
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([0,x], [0,y], color='tab:red', linewidth=4)
ax.plot(x, y,marker='o',color='r',markersize=12)
plt.text(x*.71, y*.97, f'({x:.0f},{y})', fontsize=18)
plt.xlabel('raio, mm')
plt.ylabel('Tensão de corte, MPa')
plt.grid()
plt.show()
```

É determinada a tensão máxima de corte que os veios do mecanismo podem suportar. Contudo desconhece-se qual das barras, $AB$ ou $DE$, irá suportar essa tensão máxima. Para determinar o estado de tensão de corte nas barras é necessário analisar o equilíbrio dos elementos que constituem o mecanismo.

**Equílibrio estático dos momentos de torção**

O momento torsor aplicado na secção $A$ irá ser transmitido ao longo de todo o veio $AB$ ($T_A = T_{AB}$). Este momento irá ser equilibrado por uma força tangencial na engrenagem, na secção B.

Comece-se por representar o diagrama de corpo livre das engrenagens em $B$ e $D$. O momento torsor em $A$ irá originar uma força de contacto nos dentes das engrenagens, $F$, que equilibrará o momento torsor aplicado. Por sua vez, uma força oposta (igual magnitude mas sentido oposto) será exercida sobre a engregem em $D$ (pela acção da engrenagem em $B$), provocando um momento torsor $T_{DE}$ ao longo do veio $DE$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au11/Au11_P003_i2.jpg"
width="400" height="330">

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au11/Au11_P003_i3.jpg"
width="400" height="170">

- Eq. de equílibrio para a engrenagem em $B$:

\begin{equation*}
+\circlearrowleft\,: \sum M_B = 0
~\Leftrightarrow~
T_{AB} - F\cdot r_B = 0
~\Leftrightarrow~
T_{AB} = F\cdot r_B
\end{equation*}

- Eq. de equílibrio para a engrenagem em $D$:

\begin{equation*}
+\circlearrowleft\,: \sum M_D = 0 ~\Leftrightarrow~
T_{DE} - F'\cdot r_D = 0
~\Leftrightarrow~
T_{DE} = F'\cdot r_D
\end{equation*}

As forças exercidas pelos dentes das engrenagens são iguais em amplitude mas opostas em sentido: $F=F'$.

Pela igualdade entre as força de contacto circunferencial entre dentes das engrenagens, pode-se concluir a seguinte relação:

\begin{equation*}
F = \frac{T_{AB}}{r_B} = \frac{T_{DE}}{r_C}
~\Leftrightarrow~
T_{DE}  = \frac{r_D}{r_B} T_{AB} = 1.5 \, T_{AB}
\end{equation*}

Assim sendo o momento de torção máxima irá ocorrer no veio $DE$: $T_\mathrm{max} = T_{DE} \equiv T_{D}$.

```python
TDE = Tmax
print(f'TDE = {TDE:.2f} N.mm | {TDE*1e-3:.2f} N.m')
TAB = rB/rD*TDE
print(f'TAB = {TAB:.2f} N.mm | {TAB*1e-3:.2f} N.m')
```

#### b) Determine o ângulo) de torção nesse ponto, $\phi_A$ = ?. ([_enunciatum_](#PV113))

Começe-se por notar que o ângulo de torção da barra $AB$ se escreve neste caso:

\begin{equation*}
\phi_{A/B} = \frac{T_{AB}L_{AB}}{JG} = \phi_A - \phi_B \quad(\textrm{com:}~\phi_A \gt \phi_B)
\end{equation*}

em que,
- $\phi_A$ é o ângulo de torção na extremidade A da barra,
- $\phi_B$ é o ângulo de torção na extremidade B da barra.

**Cinemática das engrenagens**

O ângulo de torção na secção $B$ (engrenagem $B$) pode relacionar-se com o  ângulo de torção na secção $D$ (engrenagem $D$) atentendo à análise cinemática das engrenagens.

Sendo a velocidade linear periférica das engrenagens igual, pode-se escrever:

\begin{equation*}
\begin{aligned}
v_B = v_D &~\Leftrightarrow~ w_Br_B = w_Dr_D \\
&~\Leftrightarrow~ \frac{d \phi_B}{dt} r_B = \frac{d \phi_D}{dt} r_D \\
&~\Leftrightarrow~ r_B \int \frac{d \phi_B}{dt} dt =  r_D \int \frac{d
\phi_D}{dt} dt \\
&~\Leftrightarrow~ \phi_B r_B = \phi_Dr_D
\end{aligned}
\end{equation*}

Geometricamente esta relação pode entender-se considerando a figura abaixo. De notar que, os comprimentos de arco $CC'$ e $CC''$ percorridos em ambas as engrenagens, no movimento de rotação, deve ser igual.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au11/Au11_P003_i4.jpg"
width="400" height="210">

\begin{equation*}
\begin{aligned}
CC' = CC'' &~\Leftrightarrow~
\phi_B \cdot r_B = \phi_D \cdot r_D
\end{aligned}
\end{equation*}

Destas equações pode concluir-se a seguinte relação dos torção de rotação das engrenagens:

\begin{equation*}
\begin{aligned}
\phi_B  = \frac{r_D}{r_B} \phi_D = 1.5 \, \phi_D
\end{aligned}
\end{equation*}

Por outro lado, sendo a extremidade em $E$ encastrada, o ângulo de torção $\phi_D$ da engrenagem $D$ é igual ao ângulo de torção do veio $DE$:

\begin{equation*}
\phi_{D/E} = \phi_{D} = \left(\frac{T L}{JG}\right)_{DE}
= \left(\frac{T_\textrm{max} L}{JG}\right)_{DE}
= \left(\frac{\tau_\textrm{amd} L}{c~G}\right)_{DE}
\end{equation*}

```python
phiD = TDE*LDE/J/G
print(f'phiD (cal. 1) = {phiD:.4f} radianos | {np.rad2deg(phiD):.3f} º')
phiD2 = Sadm*LDE/r/G
print(f'phiD (cal. 2) = {phiD2:.4f} radianos | {np.rad2deg(phiD2):.3f} º')
```

de onde se retira:

\begin{equation*}
\begin{aligned}
\phi_B  = \frac{r_D}{r_B} \left(\frac{T L}{JG}\right)_{DE}
\end{aligned}
\end{equation*}

```python
phiB = rD/rB *phiD
print(f'phiB = {phiB:.4f} radianos | {np.rad2deg(phiB):.3f} º')
```

O valor do ângulo de torção relativo da barra AB é calculado pela expressão:

\begin{equation*}
\phi_{A/B} = \phi_{A} - \phi_{B} = \left(\frac{TL}{JG}\right)_{AB}
\end{equation*}

```python
phiAB = TAB*LAB/J/G
print(f'phiAB = {phiAB:.4f} radianos | {np.rad2deg(phiAB):.3f} º')
```

Esta, pode agora ser resolvida para o ângulo de torção da secção A do veio,
resultando finalmente:

\begin{equation*}
\phi_{A} = \phi_{A/B} + \phi_{B}
= \left(\frac{TL}{JG}\right)_{AB} + \phi_{B}
= \left(\frac{TL}{JG}\right)_{AB} +
\frac{r_D}{r_B} \left(\frac{T L}{JG}\right)_{DE}
\end{equation*}

```python
phiA = TAB*LAB/J/G + phiB
print(f'phiA = {phiA:.4f} radianos | {np.rad2deg(phiA):.3f} º')
```
