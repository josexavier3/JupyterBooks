## Resolução

width="600">

O pilar de betão armado irá sofrer uma dilatação conjunta após variação de temperatura. Pelo princípio de sobreposição verificar-se que:

- o aço expande livremente pela variação de temperatura $\delta_a^T$;
- por uma questão de compatibilidade de deslocamento ficará submetido a uma força de compressão, $\delta_a^F$;
- o betão expande livremente pela variação de temperatura $\delta_b^T$;
- por uma questão de compatibilidade de deslocamento ficará submetido a uma força de tração, $\delta_b^F$;

##### Equação de equilíbrio estático

O deslocamento por dilatação térmica introduzirá uma força resultante na secção do pilar. Por outro lado, a análise do equilíbrio de forças segundo a direção vertical do pilar conduz a :

\begin{equation*}
R = \int_{A_a} \sigma\mathrm{d} A + \int_{A_b} \sigma\mathrm{d} A  = F_a +
 F_b = 0
 \quad\Leftrightarrow\quad
 F_a = - F_b
\end{equation*}

Uma vez que a resultantes das forças externas é nula, conclui-se que as forças exercidas em cada área são iguais em módulo mas opostas em sentido.

\begin{equation*}
F = |F_a| = |F_b|
\end{equation*}

A análise esquemática a figura anterior permite deduzir que a força exercida no aço é de compressão enquanto que a força de tração exercida no betão é de tração.

##### Equação de compatibilidade de deslocamentos

\begin{equation*}
\begin{aligned}
\delta_a^T - \delta_a^F &= \delta_b^T + \delta_b^F \\
\alpha_a \Delta T L - \frac{FL}{AE_a} &=
\alpha_b \Delta T L + \frac{FL}{AE_b}  \\
\therefore\quad F & = \frac{\Delta T(\alpha_a - \alpha_b) }
{\left(\frac{1}{A_aE_a} + \frac{1}{A_bE_b}\right)}
\end{aligned}
\end{equation*}

em que,

\begin{equation*}
\sigma_a = -\frac{F}{A_a}
\quad\wedge\quad \sigma_b = \frac{F}{A_b}
\end{equation*}

```python
import numpy as np

#unidades mm, MPa, N
# aço
Ea, alphaa, da = 210.e3, 11.7e-6, 22.

# betao
Eb, alphab = 25.e3, 9.9e-6

DT = 50.
Aa =  6*np.pi*da**2/4
print(f'Aa = {Aa:.2f} mm^2')
Lx, Ly, Lz = 240., 240., 1800.
Ab = Lx*Ly - Aa
print(f'Ab = {Ab:.2f} mm^2')
F = DT*(alphaa - alphab)/(1/Aa/Ea + 1/Ab/Eb)
print(f'F  = {F:.2f} N')

siga = F/Aa
print(f'siga = {siga:.2f} MPa (tração)')
sigb = F/Ab
print(f'siga = {sigb:.2f} MPa (compressão)')
```
