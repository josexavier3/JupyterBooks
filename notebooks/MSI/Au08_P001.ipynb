## Problema 1 <a name="PIV81"></a>

As barras $AB$, $CD$, $BC$ e $AD$ da estrutura em treliça indicada na figura são cilíndricas e feitas de aço ($E= 210$ GPa). O diâmetro das barras $A$B e $CD$ é de 30 mm e o diâmetro das barras $BC$ e $AD$ é de 20 mm. Quando o esticador está apertado, a barra diagonal $A$C fica tracionada. Sabendo que $h = 1,4$ m, determine a força máxima admissível em $AC$ tal que os alongamentos nas barras $AB$ e $CD$ não excedam 1 mm.

## Resolução

Se a estrutura está em equilíbrio estático, isso significa que também os seus elementos (barras e nós) estão em equilíbrio. Considere-se portanto a análise do equilíbrio de forças que atuam no nó C (diagrama de corpo livre do nó C):

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au08/Au08_P001_i2.jpg"
width="350" height="140">

A análise da força que atua na barra $CD$ vem:

\begin{equation*}
F_\mathrm{CD} = \delta_\mathrm{CD} \frac{EA_\mathrm{CD}}{L_\mathrm{CD}}~\textrm{(N)}~;
\quad \delta_\mathrm{CD} \leq 1~\textrm{(mm)}
\end{equation*}

em que
\begin{equation*}
A_\mathrm{CD} = \frac{\pi d_\mathrm{CD}^2}{4}~\textrm{(mm$^2$)}~;~L_\mathrm{CD} = h
~\textrm{(mm)}\end{equation*}

pelo que resulta:

\begin{equation*}
F_\mathrm{CD} \leq \delta_\mathrm{CD} \frac{E \pi d_\mathrm{CD}^2}{4h}
~\textrm{(N)}
\end{equation*}

Da análise do equilíbrio no nó $C$ : $\sum F_y = 0$ resulta

\begin{equation*}
F_\mathrm{CD} - F_\mathrm{AC}\cos\beta=0
\end{equation*}

Da análise geométrica da orientação das forças no nó $C$ resultam as seguintes relações trigonométricas:

\begin{equation*}
\sin\beta = \frac{1}{\sqrt{1+h^2}}~;~
\cos\beta = \frac{h}{\sqrt{1+h^2}}
\end{equation*}

Pelo que,

\begin{equation*}
\begin{aligned}
F_\mathrm{AC} = \frac{F_\mathrm{CD}}{\cos\beta}  = F_\mathrm{CD} \frac{\sqrt{1+h^2}}{h}
\leq \delta_\mathrm{CD} \frac{E \pi d_\mathrm{CD}^2}{4h} \sqrt{1+h^2}
\end{aligned}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy import diff, Derivative
from sympy.solvers import solve
from sympy import symbols
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
from matplotlib import rcParams, rc
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 18
rc('text', usetex=True)

E = 210*1e3 # unit: MPa
print('Propriedades do material: ---------------')
print('     E (MPa) = ',E)

# unidade: mm
dAB, dCD, dBC, dAD = 30, 30, 20, 20

h = 1.4e3 # unidade: mm

LAB = h   # unidade: mm
LCD = 1e3 # unidade: mm

alongAB = 1 # menor ou igual a 1 mm
alongCD = 1 # < =  1 mm

print('Versão 1 ---------------------')
cosb = LAB/np.sqrt(LCD**2+LAB**2)
print(f'beta = {np.rad2deg(np.arccos(cosb)):.3f} º')
# ou: print(np.rad2deg(np.arctan(1/h)))

FAC1 = float(alongCD*E*np.pi*dCD**2*np.sqrt(LCD**2+h**2)/(4*h**2))
print(f'FAC \leq = {FAC1*1e-3:.2f} kN')
```

Em alternativa, considere-se a seguite resolução:

Equilíbrio estático da estrutura isostática é traduzido pelas seguintes equações esclares:

\begin{equation*}
\begin{cases}
\sum F_x &= 0 \\[1ex]
\sum F_y &= 0 \\[1ex]
\sum M_A &= 0 \\
\end{cases}~\Leftrightarrow~
\begin{cases}
R_{Ax} &= 0 \\[1ex]
R_{Ay} + R_{Dy} &= 0 \\[1ex]
R_{Dy}&= 0 \\
\end{cases}
\end{equation*}

Por outro lado, os esforços nas barras verificam as seguintes relações:

\begin{equation*}
F_{AD} = F_{BC}~;~ \quad F_{AB} = F_{CD}
\end{equation*}

Vem ainda:

\begin{equation*}
F_{AB} = F_{AC} \cos\beta
\end{equation*}

Relação entre força e deslocamento:

\begin{equation*}
\delta_i = \frac{F_iL_i}{A_iE_i}
\end{equation*}

De onde se resolve,

\begin{equation*}
\begin{aligned}
\delta_{AB} = \frac{F_{AB}L_{AB}}{A_{AB}E} &\le 1\\
F_{AC} \cos\beta \frac{L_{AB}}{A_{AB}E} &\le 1\\
F_{AC} &\le \frac{A_{AB}E}{L_{AB}\cos\beta}
\end{aligned}
\end{equation*}

```python
print('Versão 2 ---------------------')
ang = np.arctan(LCD/LAB)
print(f'ângulo entre AB e diagonal AC: {np.rad2deg(ang):.3f} deg')

AAB = np.pi*(dAB/2)**2# área de um círculo
print(f' A:AB = {AAB:.3f} mm^2')
FAC2 = (AAB*E)/h/np.cos(ang)
print(f' FAC  <= {FAC2*1e-3:.2f} kN')
```
