## Lei de Hooke. Parte I
## Problema 4 <a name="PIII64"></a>

Um reservatório cilíndrico com 600 mm de diâmetro interior e 10 mm de espessura de parede, sujeito a pressão interna, tem montado na sua superfície dois extensómetros perpendiculares entre si. O extensómetro transversal regista uma extensão de 255$\mu\varepsilon$ e o outro 60$\mu\varepsilon$. Desconhece-se o material do cilindro mas de um ensaio de torção determinou-se experimentalmente o módulo de rigidez $G = $ 80 GPa.

Assim, com os dados de que dispõe calcule:

[a)](#A) A tensão de corte máxima no plano da parede do cilindro.

[b)](#B) A pressão interna.

[c)](#C) As tensões principais.

[d)](#D) A tensão de corte máxima (considere os três planos de tensões de corte máxima) na parede do cilindro.


## Resolução

### a) A tensão de corte máxima no plano da parede do cilindro.<a name="A"></a>
([enunciatum](#PIII64))

O pressuposto de reservatório parede fina assume a validade da seguinte relação: (diâmetro / espessura = d / t > 20):

```python
# dimensões do reservatório
d = 600
t = 10
r = d/2
print(' d/t = ',d/t)
```

A validade desta relação permite analisar a parede como uma superfície e, consequentemente, usar a equação de Young-Laplace para estimar a tensão circunferêncial criada por uma pressão interna no reservatório de pressão cilíndrica de paredes finas. Ou seja, como as paredes oferecem pouca resistência à flexão pode-se supor que os esforços internos que atuam na superfície (parede) sejam tangentes a esta. Logo pode assumir-se um estado plano de tensão na superfície. Considerando um sistema de coordenadas $S(0,x,y)$, este é referencial principal sendo as tensões circunferencial ($\sigma_{\theta}$), tensões longitudinais ($\sigma_{l}$) e tensões radiais ($\sigma_{r}$) tensões principais ($\tau_{xy} = 0$):

\begin{equation*}
\begin{cases}
\sigma_{\theta} &\equiv \sigma_{yy} = \displaystyle\frac{Pr}{t} \\
\sigma_{l} &\equiv \sigma_{xx} = \displaystyle\frac{Pr}{2t} \\
\sigma_{r} &\equiv \sigma_{zz} = -\displaystyle\frac{P}{2} \\
\end{cases}
\end{equation*}

Em que, $P$ é a pressão interna, $t$ é a espessura da parede, e $r$ é o raio médio do cilindro.

A tensão circunferêncial pode ser deduzida da relação:

\begin{equation*}
\begin{aligned}
\sum F_z &= 0 \\
\sigma_\theta \mathrm{d}A - P \mathrm{d}A &= 0 \\
\sigma_\theta (2bt) - P (2rb) &= 0 \\
\sigma_\theta &= \frac{Pr}{t} \\
\end{aligned}
\end{equation*}

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au06/Au06_P004_i1.jpg"
width="350" height="200">

Portanto as extensões lineares $\varepsilon_{0^\circ}$ e $\varepsilon_{90^\circ}$ são extensões principais:

\begin{equation*}
\underline{\underline{\varepsilon}}_\mathrm{P} =
\begin{bmatrix}
\varepsilon_{0^\circ} & 0 \\
0 & \varepsilon_{90^\circ} \\
\end{bmatrix} =
\begin{bmatrix}
60 & 0 \\
0 & 255 \\
\end{bmatrix}\quad(\mu\varepsilon)
\end{equation*}

Pela análise da expressão de transformação do tensor das deformações ou pelo círculo de Mohrh têm-se:

\begin{equation*}
\frac{1}{2}\gamma_\mathrm{max} = \frac{1}{2}(\varepsilon_{xx} - \varepsilon_{yy})
\end{equation*}

Pela lei de Hooke para um material isotrópico e homogéneo:

\begin{equation*}
\gamma_{xy} = \frac{\tau_{xy}}{G} \Leftrightarrow \tau_\mathrm{max} = \gamma_\mathrm{max} G
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=3)

ea = 60*1e-6
eb = 255*1e-6
mDP = np.array([[ea,0],[0,eb]])
print('mDP = \n',mDP)

eI  = mDP[1,1]
eII = mDP[0,0]

gxyP = (eI - eII)
print('\ngxyP (=gmax) = ','{:.5e}'.format(gxyP),' [radianos]')
print('gxyP (=gmax) = ','{:.1f}'.format(gxyP*1e6),' [micro-radianos]')

G = 80*1e3 # unit: MPa
print('Propriedades do material: ---------------')
print('     G (MPa) = ','{:.2f}'.format(G))

tmax = gxyP*G
print('sxyP (=smax) = ','{:.1f}'.format(tmax),' MPa')
```

#### b) A pressão interna.<a name="B"></a>

([enunciatum](#PIII64))

Pela análise da lei de tranformação do tensor das tensões ou pelo círculo de Mohr:

\begin{equation*}
\sigma_\mathrm{min} \equiv \sigma_{xx} = \frac{Pr}{2t} \quad\textrm{e}\quad
\sigma_\mathrm{max} \equiv \sigma_{yy} = \frac{Pr}{t}
\tau_{xy} = 0
\end{equation*}

\begin{equation*}
\tau_\mathrm{max}
= \frac{1}{2} (\sigma_\mathrm{max} - \sigma_\mathrm{min})
= \frac{1}{2} \left(\frac{Pr}{t} - \frac{Pr}{2t} \right) = \frac{Pr}{4t}
\Leftrightarrow P = \frac{4t \tau_\mathrm{max}}{r}
\end{equation*}

```python
# pressão interna no reservatório
P = 4*t*tmax/r
print('P = ','{:.3f}'.format(P),' MPa')
```

#### c) As tensões principais.<a name="C"></a>

([enunciatum](#PIII64))

Pela análise da lei de tranformação do tensor das tensões ou pelo círculo de Mohr:

\begin{equation*}
\begin{aligned}
\sigma_\mathrm{min} \equiv \sigma_{xx} \equiv \sigma_{l} \equiv \sigma_\mathrm{II} = \frac{Pr}{2t} \\
\sigma_\mathrm{max} \equiv \sigma_{yy} \equiv \sigma_{\theta} \equiv \sigma_\mathrm{I}= \frac{Pr}{t}
\end{aligned}
\end{equation*}

\begin{equation*}
\underline{\underline{\sigma}}_\mathrm{P} = [\sigma]_\mathrm{P} =
\begin{bmatrix}
\sigma_\mathrm{I} & 0 & 0 \\
0 & \sigma_\mathrm{II} & 0 \\
0 & 0 & 0 \\
\end{bmatrix}
\end{equation*}

```python
sxx = P*r/2/t
print('Sxx = ','{:.3f}'.format(sxx),' MPa')
syy = P*r/t
print('Syy = ','{:.3f}'.format(syy),' MPa')
```

#### d) A tensão de corte máxima (considere os três planos de tensões de corte máxima) na parede do cilindro.<a name="D"></a>

([enunciatum](#PIII64))

Tensão de corte máxima entre:

\begin{equation*}
\begin{cases}
\tau_\mathrm{max}^{yz} &= \frac{1}{2}(\sigma_\mathrm{I} - \sigma_\mathrm{III}) \\
\tau_\mathrm{max}^{xy} &= \frac{1}{2}(\sigma_\mathrm{I} - \sigma_\mathrm{II}) \\
\tau_\mathrm{max}^{xz} &= \frac{1}{2}(\sigma_\mathrm{II} - \sigma_\mathrm{III}) \\
\end{cases}
\end{equation*}

```python
sI = syy
sII = sxx
sIII = 0

tcorteyz = 0.5*(sI - sIII)
print('tens. corte max : yz =','{:.2f}'.format(tcorteyz),' MPa')
tcortexy = 0.5*(sI - sII)
print('tens. corte max : xy =','{:.2f}'.format(tcortexy),' MPa')
tcortexz = 0.5*(sII - sIII)
print('tens. corte max : yz =','{:.2f}'.format(tcortexz),' MPa')
```

Pelo que $\tau_\mathrm{max} = 31.2$ MPa.

```python
def circleradius(S11,S22,S12):
    center = [(S11 + S22)/2.0, 0.0]
    radius = np.sqrt((S11 - S22)**2/4.0 + S12**2)
    return center, radius

limits = np.array([[-5,70],[-40,40]])

center1, radius1  = circleradius(sI,sIII,0)
center2, radius2  = circleradius(sI,sII,0)
center3, radius3  = circleradius(sII,sIII,0)

circ1 = plt.Circle((center1[0],0), radius1, facecolor='b', lw=3,edgecolor='b')
circ2 = plt.Circle((center2[0],0), radius2, facecolor='r', lw=3,edgecolor='r')
circ3 = plt.Circle((center3[0],0), radius3, facecolor='g', lw=3,edgecolor='g')

plt.axis('image'); ax = plt.gca()

plt.gcf().gca().add_artist(circ1)
plt.gcf().gca().add_artist(circ2)
plt.gcf().gca().add_artist(circ3)

plt.xlabel(r"$\sigma$", size=18)
plt.ylabel(r"$\tau$", size=18)

plt.text(1.*sI, 0, r'$\sigma_\mathrm{I}$')
plt.text(1.*sII, 0, r'$\sigma_\mathrm{II}$')
plt.text(1.*sIII, 0, r'$\sigma_\mathrm{III}$')

plt.xlim(limits[0,0],limits[0,1])
plt.ylim(limits[1,0],limits[1,1])

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.show()
```
