## Resolução

##### Deformação angular

A deformação de um corpo traduz-se pela variação das distâncias entre os seus pontos materiais. Na prática, esta
deformação linear é ainda acompanhada por uma alteração do ângulo entre duas direções definidas num ponto $P$ genérico
da configuração inicial ou de referência.

Considere-se uma região elementar em torno do ponto $P$. Tome-se neste pontos dois vetores
$\overrightarrow{V}_1 \equiv \overrightarrow{PQ}$ e $\overrightarrow{V}_2 \equiv \overrightarrow{PR}$,
ambos com origem no ponto $P$ e com direções iniciais definidas pelos versores
$\vec{n}_1 \equiv \vec{n}_{PQ} = n_{1x} \vec{\imath} + n_{1y} \vec{\jmath} + n_{1z} \vec{k}$ e
$\vec{n}_2 \equiv \vec{n}_{PR} = n_{2x} \vec{\imath} + n_{2y} \vec{\jmath} + n_{2z} \vec{k}$. Com a deformação do corpo as fibras $\vec{V}_1$ e $\vec{V}_2$ transformar-se-ão nos
vetores $\vec{V}\!'_1$ e $\vec{V}\!'_2$, respetivamente. Em geral a deformação envolve não apenas a variação do
comprimento das fibras $\vec{V}_1$ e $\vec{V}_2$ mas também a variação do ângulo entre eles, designada por deformação
angular. Seja $\theta$ o ângulo inicial formado entre $\vec{V}_1$ e $\vec{V}_2$ antes da deformação e seja $\theta'$ o
ângulo final formado entre $\vec{V}\!'_1$ e $\vec{V}\!'_2$ após deformação. Por definição do produto escalar entre
vetores resulta diretamente:

\begin{equation}
\cos\theta = \frac{\vec{V}_1 \cdot \vec{V}_2}{||\vec{V}_1||~||\vec{V}_2||}
= \frac{\vec{V}_1 \cdot \vec{V}_2}{V_1V_2} = n_{1x}n_{2x} + n_{1y}n_{2y} + n_{1z}n_{2z}
\end{equation}

e,

\begin{equation}
\cos\theta' = \frac{\vec{V}'_1 \cdot \vec{V}'_2}{||\vec{V}'_1||~||\vec{V}'_2||}
= \frac{\vec{V}_1 \cdot \vec{V}'_2}{V_1'V_2'} == n'_{1x}n'_{2x} + n'_{1y}n'_{2y} + n'_{1z}n'_{2z}
\end{equation}

Uma vez que $\vec{V}' = \vec{V} + \delta\vec{V}$ e $\vec{V}' = \vec{V} (1 + \varepsilon)$, a última expressão pode ser
reescriva por:

\begin{equation}
\begin{aligned}
\cos\theta' &= \frac{(\vec{V}\!'_1 + \delta\vec{V}\!'_1) \cdot (\vec{V}\!'_2 + \delta\vec{V}\!'_2)}
{V_1V_2(1 + \varepsilon_1)(1 + \varepsilon_2)} \\
&= \frac{\cos\theta + \vec{n}_1\cdot\vec{A}_2 + \vec{n}_2\cdot\vec{A}_1 + \vec{A}_1\cdot\vec{A}_2 }
{(1 + \varepsilon_1)(1 + \varepsilon_2)} \\
&= \frac{\cos\theta + \{n\}_1^{T}[A]\{n\}_2 + \{n\}_2^{T}[A]\{n\}_1 + \{n\}_1^{T}[A]^T[A]\{n\}_2}
{(1 + \varepsilon_1)(1 + \varepsilon_2)} \\
\end{aligned}
\end{equation}

em que a matriz $[A]$ é a matriz do gradiente dos deslocamentos definida por:

\begin{equation}
[A] \equiv \mathrm{grad} \vec{u} \equiv \vec{\nabla} \vec{u} =
\begin{bmatrix}
\displaystyle\left(\frac{\partial u}{\partial x}\right)_P &
\displaystyle\left(\frac{\partial u}{\partial y}\right)_P &
\displaystyle\left(\frac{\partial u}{\partial z}\right)_P \\
\displaystyle\left(\frac{\partial v}{\partial x}\right)_P &
\displaystyle\left(\frac{\partial v}{\partial y}\right)_P &
\displaystyle\left(\frac{\partial v}{\partial z}\right)_P \\
\displaystyle\left(\frac{\partial w}{\partial x}\right)_P &
\displaystyle\left(\frac{\partial w}{\partial y}\right)_P &
\displaystyle\left(\frac{\partial w}{\partial z}\right)_P
\end{bmatrix}
\end{equation}

Como, $\{n\}_2^{T}[A]\{n\}_1 = (\{n\}_2^{T}[A]\{n\}_1)^T = \{n\}_1^{T}[A]^T\{n\}_2$, conclui-se que:

\begin{equation}
\begin{aligned}
\cos\theta' &=
\frac{\cos\theta}{(1 + \varepsilon_1)(1 + \varepsilon_2)} +
\frac{2}{(1 + \varepsilon_1)(1 + \varepsilon_2)}\{n\}_1^{T}[E]\{n\}_2
\end{aligned}
\end{equation}

em que a matriz $[E]$ é a matriz das deformações de Lagrange:

\begin{equation}
\begin{aligned}[t]
[E] &= \frac{1}{2}[A] + \frac{1}{2}[A]^\mathrm{T} + \frac{1}{2}[A]^\mathrm{T}[A] \\
& =
\begin{bmatrix}
E_{xx} & G_{xy} & G_{xz} \\
G_{xy} & E_{yy} & G_{yz} \\
G_{xz} & G_{yz} & E_{zz}
\end{bmatrix}
\end{aligned}
\end{equation}

com,

\begin{equation}
\begin{aligned}
E_{xx} & = \frac{\partial u}{\partial x} + \frac{1}{2}\left[ \left(\frac{\partial u}{\partial x}\right)^2 +
\left(\frac{\partial v}{\partial x}\right)^2 + \left(\frac{\partial w}{\partial x}\right)^2\right]\\
E_{yy} & = \frac{\partial u}{\partial x} + \frac{1}{2}\left[ \left(\frac{\partial u}{\partial y}\right)^2 +
\left(\frac{\partial v}{\partial y}\right)^2 + \left(\frac{\partial w}{\partial y}\right)^2\right] \\
E_{zz} & = \frac{\partial u}{\partial x} + \frac{1}{2}\left[ \left(\frac{\partial u}{\partial z}\right)^2 +
\left(\frac{\partial v}{\partial z}\right)^2 + \left(\frac{\partial w}{\partial z}\right)^2\right]\\
G_{xy} & = \frac{1}{2}\left(\frac{\partial u}{\partial y} + \frac{\partial v}{\partial x}\right) +
\frac{1}{2} \left( \frac{\partial u}{\partial x}\frac{\partial u}{\partial y} +
\frac{\partial v}{\partial x}\frac{\partial v}{\partial y} +
\frac{\partial w}{\partial x}\frac{\partial w}{\partial y} \right) \\
G_{xz} & = \frac{1}{2}\left(\frac{\partial u}{\partial z}+\frac{\partial w}{\partial x}\right) +
\frac{1}{2} \left( \frac{\partial u}{\partial x}\frac{\partial u}{\partial z} +
\frac{\partial v}{\partial x}\frac{\partial v}{\partial z} +
\frac{\partial w}{\partial x}\frac{\partial w}{\partial z} \right)\\
G_{yz} & = \frac{1}{2}\left(\frac{\partial v}{\partial z}+\frac{\partial w}{\partial y}\right) +
\frac{1}{2} \left( \frac{\partial u}{\partial y}\frac{\partial u}{\partial z} +
\frac{\partial v}{\partial y}\frac{\partial v}{\partial z} +
\frac{\partial w}{\partial y}\frac{\partial w}{\partial z} \right)\\
\end{aligned}
\end{equation}

Para pequenas deformações têm-se $[E] \approx [\varepsilon] = 1/2([A]+[A]^T)$ pelo que a deformação de Lagrande é
infinitesimal e confunde-se com a deformação linear: $[E] \approx \varepsilon$. Além disso,
$(1+\varepsilon_1)(1+\varepsilon_2) \approx 1 $, pelo que resulta:

\begin{equation}
\begin{aligned}
\cos\theta' &=
\cos\theta +
2\{n\}_1^{T}[D]\{n\}_2
\end{aligned}
\end{equation}

em que,

\begin{equation}
\begin{aligned}
\underline{\underline{\varepsilon}} \equiv [\varepsilon] \equiv \varepsilon_{ij} &=
\begin{bmatrix}
\varepsilon_{xx} & \varepsilon_{xy} & \varepsilon_{xz} \\
\varepsilon_{xy}& \varepsilon_{yy} &  \varepsilon_{yz} \\
\varepsilon_{xz} & \varepsilon_{yz} & \varepsilon_{zz} \\
\end{bmatrix} =
\begin{bmatrix}
\varepsilon_{xx} & \gamma_{xy}/2 & \gamma_{xz}/2 \\
\gamma_{yx}/2 & \varepsilon_{yy} & \gamma_{yz}/2 \\
\gamma_{zx}/2 & \gamma_{zy}/2 & \varepsilon_{zz} \\
\end{bmatrix} \\
&=
\begin{bmatrix}
\displaystyle\frac{\partial u}{\partial x}
& \displaystyle\frac{1}{2}\left( \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} \right)
& \displaystyle\frac{1}{2}\left( \frac{\partial u}{\partial z} + \frac{\partial w}{\partial x} \right)\\
\displaystyle\frac{1}{2}\left( \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} \right)
&  \displaystyle\frac{\partial v}{\partial y}
& \displaystyle\frac{1}{2}\left( \frac{\partial v}{\partial z} + \frac{\partial w}{\partial y} \right) \\
\displaystyle\frac{1}{2}\left( \frac{\partial w}{\partial z} + \frac{\partial w}{\partial x} \right)
& \displaystyle\frac{1}{2}\left( \frac{\partial v}{\partial z} + \frac{\partial w}{\partial y} \right)
& \displaystyle\frac{\partial w}{\partial z} \\
\end{bmatrix}
\end{aligned}
\end{equation}

Para o caso de um ângulo reto entre os dois segmentos de linha infinitesimais, a equação anterior simplifica-se para:

\begin{equation}
\begin{aligned}
\cos\theta' &=
2\{n\}_1^{T}[D]\{n\}_2
\end{aligned}
\end{equation}

Analize-se com mais detalhe o caso particular para o qual o ângulo inicial $\theta$ é um ângulo reto. Neste caso,
chama-se deformação angular no ponto $P$ e entre as duas direções dos versores ortogonais $\vec{n}_1$ e $\vec{n}_2$ ao
ângulo assim  definido:

\begin{equation}
\gamma_{12}(P,\vec{n}_1,\vec{n}_2) = \frac{\pi}{2} - \theta'
\end{equation}

da expressão de $\cos\theta'$ é possível concluir que (notas: $\cos\theta \equiv \cos90^\circ = 0$, e
$\cos\theta' = \cos(\pi/2 - \gamma_{12}) = \sin\gamma_{12} $):

\begin{equation}
\sin\gamma_{12} = \frac{2}{(1 + \varepsilon_1)(1 + \varepsilon_2)}\{n\}_1^{T}[E]\{n\}_2
\end{equation}

sendo $l_1,m_1,n_1$ as componentes do versor $\vec{n}_1$ e $l_2,m_2,n_2$ as componentes do versor $\vec{n}_2$ no sistema
de coordenadas $S_0(0,\vec{\imath},\vec{\jmath},\vec{k})$

\begin{equation}
\sin\gamma_{12} = \frac{2}{(1 + \varepsilon_1)(1 + \varepsilon_2)}
[E_{xx}l_1l_2 + E_{yy}m_1m_2 + E_{zz}n_1n_2 + G_{xy}(l_1m_2+l_2m_1) + G_{xz}(l_1n_2+l_2n_1) + G_{yz}(m_1n_2 + m_2n_1)]
\end{equation}

Esta expressão determina a deformação angular $\gamma_{12}$ no ponto $P$ e entre as direções $\vec{n}_1$ e $\vec{n}_2$,
à custa dos elementos da matriz das deformações de Lagrange.

Para o caso das pequenas deformações, a expressão da deformação angular simplifica-se uma vez que:

\begin{equation}
(1+\varepsilon_1)(1+\varepsilon_2) \approx 1,
\end{equation}

pelo que se pode supor que a deformação angular é independente das
deformações lineares segundo $\vec{n}_1$ e $\vec{n}_2$. Por outro lado, assumindo um ângulo $\gamma_{12}$ infinitesimal,
resulta:

\begin{equation}
\sin\gamma_{12} \approx \gamma_{12}
\end{equation}

Assim, sendo resulta:

\begin{equation}
\begin{aligned}
\gamma_{12} &= 2 \{n\}_1^{T}[D]\{n\}_2 \\
& = 2\varepsilon_{xx}l_1l_2 + 2\varepsilon_{yy}m_1m_2 + 2\varepsilon_{zz}n_1n_2 +
\gamma_{xy}(l_1m_2 + l_2m_1) + \gamma_{xz}(l_1n_2 + l_2n_1) + \gamma_{yz}(m_1n_2 + m_2n_1)
\end{aligned}
\end{equation}

em que $(l_i,m_i,n_i)~(i = 1,2)$ são as componentes do versos $\vec{n}_i~(i=1,2)$. De acordo com a definição de
deformação angular os versores $\vec{n}_1$ e $\vec{n}_2$ são ortogonais.

Os elementos não diagonais da matriz das deformações $[D]$ podem ser obtidos da expressão genérica anterior
($\gamma_{12}$) considerando os eixos de um sistema de coordenadas ortonormado e direto. Considere-se o versor
$\vec{\imath}$ e $\vec{\jmath}$ do sistema de coordenadas $S_0$. A título de exemplo, considerando o versos $\vec{n}_1$
coincidente com $\vec{\imath}$ e o versor $\vec{n}_2$ coincidente com $\vec{\jmath}$; as suas componentes em $S_0$ vêm:

\begin{equation}
\begin{aligned}
\vec{n}_1 &: l_1 = 1, m_1 = 0, n_1 = 0 \\
\vec{n}_2 &: l_2 = 0, m_2 = 1, n_2 = 0 \\
\end{aligned}
\end{equation}

pelo que se conclui: $\gamma_{12}(P,\vec{n}_1,\vec{n}_2) \equiv \gamma_{12}(P,\vec{\imath},\vec{\jmath}) = \gamma_{xy}$.

```python
def versor(l,m,n): return np.array([l,m,n])

# versor 1: PQ
l1 = 1/np.sqrt(3)
m1 = 1/np.sqrt(3)
n1 = 1/np.sqrt(3)

# versor 2: PR
l2 = 1/np.sqrt(2)
m2 = 1/np.sqrt(2)
n2 = 0

print("visualizacao da orientação espacial dos versores n1 e n2 :")
v1 = np.array([[0, 0, 0, l1, m1, n1]])
X1, Y1, Z1, U1, V1, W1 = zip(*v1)
v2 = np.array([[0, 0, 0, l2, m2, n2]])
X2, Y2, Z2, U2, V2, W2 = zip(*v2)

plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=18)
fig, ax = plt.subplots(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X1, Y1, Z1, U1, V1, W1)
ax.quiver(X2, Y2, Z2, U2, V2, W2)
ax.set_xlim([0, .8])
ax.set_ylim([0, .8])
ax.set_zlim([0, .8])
ax.view_init(20, 20)
plt.show()
```

O ângulo inicial entre os segmentos é dado pela definição de produto escalar:

\begin{equation}
\cos\theta = \frac{\vec{n}_{PQ} \cdot \vec{n}_{PR}}{||\vec{n}_{PQ}||~||\vec{n}_{PR}||}
= \vec{n}_{PQ} \cdot \vec{n}_{PR} = \underline{n}_{PQ}^T \underline{n}_{PR}
=
\begin{bmatrix}
l_1 & m_1 & n_1
\end{bmatrix}
\begin{bmatrix}
l_2 \\ m_2 \\ n_2
\end{bmatrix} = l_1l_2 + m_1m_2 + n_1n_2
\end{equation}

```python
print('Ângulo entre segmentos PQ e PR antes de deformação')

n1 = versor(l1,m1,n1)
print('Versor n1:\n',n1)
n2 = versor(l2,m2,n2)
print('Versor n2:\n',n2)

costhetaI = np.dot(n1,n2)
thetaI = np.arccos(float(costhetaI))
thetaIDeg = np.rad2deg(thetaI)
print('cos theta (inicial) = ','{:.5f}'.format(costhetaI))
print('    theta (inicial) = ','{:.4f}'.format(thetaIDeg),' deg')
```

Visualização do campo de deslocamentos.

```python
u, v, w, x, y, z, k = symbols('u v w x y z k')

def u(x,y,k): return k*(x**2 + y)
def v(y,z,k): return k*(y+z)
def w(x,z,k): return k*(x**2+2*z**2)

xgrid = np.arange(-5.,5,.5); ygrid = xgrid; zgrid = xgrid
Xu,Yu = np.meshgrid(xgrid, ygrid) # grid of point
Yv,Zv = np.meshgrid(ygrid, zgrid)
Xw,Zw = np.meshgrid(xgrid, zgrid)

def PlotFig(X,Y,Z,stg):
    plt.figure(111)
    colorbar(imshow(Z,cmap=cm.hot))
    title(stg), plt.axis('off')
    plt.show()

k = 1e-3
PlotFig(Xu,Yu,u(Xu,Yu,k),['u = ',u(x,y,k)])
PlotFig(Yv,Zv,v(Yv,Zv,k),['v = ',v(y,z,k)])
PlotFig(Xw,Zw,w(Xw,Zw,k),['w = ',w(x,z,k)])
```

As componentes de deformação - deformação de Lagrange $[E]$ ou deformação infinitesimal $[\varepsilon]$ - no ponto $P$
podem ser obtidas a partir do campo de deslocamentos.

A dirtorção angular em função das deformações de Lagrange:

\begin{equation}
\begin{aligned}
\cos\theta' &=
\frac{\cos\theta}{(1 + \varepsilon_1)(1 + \varepsilon_2)} +
\frac{2}{(1 + \varepsilon_1)(1 + \varepsilon_2)}\{n\}_1^{T}[E]\{n\}_2
\end{aligned}
\end{equation}

```python
k = symbols('k')

u1 = u(x,y,k)
v1 = v(y,z,k)
w1 = w(y,z,k)

# componentes cartesianas da matriz das deformações de Green-Lagrange
Exx = diff(u1, x) + 0.5*(diff(u1, x)**2 + diff(v1, x)**2 + diff(w1, x)**2)
print('\nExx :',Exx)
Eyy = diff(v1, y) + 0.5*(diff(u1, y)**2 + diff(v1, y)**2 + diff(w1, y)**2)
print('Eyy :',Eyy)
Ezz = diff(w1, z) + 0.5*(diff(u1, z)**2 + diff(v1, z)**2 + diff(w1, z)**2)
print('Ezz :',Ezz)
Gxy = 0.5*(diff(u1, y) + diff(v1, x)) + 0.5*(diff(u1,x)*diff(u1,y) + diff(v1,x)*diff(v1,y) + diff(w1,x)*diff(w1,y))
print('Gxy :',Gxy)
Gxz = 0.5*(diff(u1, z) + diff(w1, x)) + 0.5*(diff(u1,x)*diff(u1,z) + diff(v1,x)*diff(v1,z) + diff(w1,x)*diff(w1,z))
print('Gxz :',Gxz)
Gyz = 0.5*(diff(v1, z) + diff(w1, y)) + 0.5*(diff(u1,y)*diff(u1,z) + diff(v1,y)*diff(v1,z) + diff(w1,y)*diff(w1,z))
print('Gyz :',Gyz)

def mE(Exx,Gxy,Gxz,Eyy,Gyz,Ezz):
    return np.array([[Exx,Gxy,Gxz],[Gxy,Eyy,Gyz],[Gxz,Gyz,Ezz]])

mE0 = mE(Exx,Gxy,Gxz,Eyy,Gyz,Ezz)
print('\nMatriz das deformações, [E] :')
print('\n[E] = \n',mE0)

Px0 = 2
Py0 = 2
Pz0 = 3

print('\nPonto =',Px0,Py0,Pz0)

Exx = Exx.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
Eyy = Eyy.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
Ezz = Ezz.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
Gxy = Gxy.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
Gxz = Gxz.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
Gyz = Gyz.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})

mEP = mE(Exx,Gxy,Gxz,Eyy,Gyz,Ezz)
np.set_printoptions(formatter={'float': '{: 0.6f}'.format})
print('\n[E(P)] = \n',mEP)

ePQ = np.dot(n1,np.dot(mEP,n1))
print('\nePQ(P) = ',ePQ)

ePR = np.dot(n2,np.dot(mEP,n2))
print(' ePR(P) = ',ePR)

costhetaLag = costhetaI/((1+ePQ)*(1+ePR)) +2/((1+ePQ)*(1+ePR))*np.dot(n1,np.dot(mEP,n2))
thetaLag = np.arccos(float(costhetaLag))
thetaLagDeg = np.rad2deg(thetaLag)
DthetaLag = thetaLagDeg- thetaIDeg
print('\n Via deformação de Lagrange : \n ')
print('\n cos theta final = ',costhetaLag)
print('     theta (final) = ','{:.4f}'.format(thetaLagDeg),' deg')
print(' Variação do ângulo (f-i) = ','{:.4f}'.format(DthetaLag),' deg')
```

A dirtorção angular em função das deformações infinitesimais:

\begin{equation}
\begin{aligned}
\cos\theta' &=
\cos\theta +
2\{n\}_1^{T}[D]\{n\}_2
\end{aligned}
\end{equation}

```python
# componentes cartesianas da matriz das deformações
exx = diff(u1, x)
print('\nexx = du/dx :',exx)
eyy = diff(v1, y)
print('eyy = dv/dy :',eyy)
ezz = diff(w1, z)
print('ezz = dw/dz :',ezz)
exy  = 0.5*(diff(u1, y) + diff(v1, x) )
print('exy = 1/2 gxy = 1/2 (du/dy + dv/dx )  :',exy)
exz  = 0.5*(diff(u1, z) + diff(w1, x) )
print('exz = 1/2 gxz = 1/2 (du/dz + dw/dx )  :',exz)
eyz  = 0.5*(diff(v1, z) + diff(w1, y) )
print('eyz = 1/2 gxy = 1/2 (dv/dz + dw/dy )  :',eyz)

print('\nMatriz das deformações, [D] :')
def mD(exx,exy,exz,eyy,eyz,ezz):
    return np.array([[exx,exy,exz],[exy,eyy,eyz],[exz,eyz,ezz]])

mD0 = mD(exx,exy,exz,eyy,eyz,ezz)
print('[D] =\n',mD0)

exx = exx.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
eyy = eyy.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
ezz = ezz.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
exy = exy.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
exz = exz.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})
eyz = eyz.subs({x:Px0,y:Py0,z:Pz0,k:1e-3})

mDP = mD(exx,exy,exz,eyy,eyz,ezz)
print('\n[D(P)] = \n',mDP)

costhetaInf = costhetaI + 2 *np.dot(n1,np.dot(mDP,n2))
thetaInf = np.arccos(float(costhetaInf))
thetaInfDeg = np.rad2deg(thetaInf)
DthetaInf = thetaInfDeg- thetaIDeg
print('\n Via deformações infinitesimal : \n ')
print('          cos theta final = ',costhetaInf)
print('            theta (final) = ','{:.4f}'.format(thetaInfDeg),' deg')
print(' Variação do ângulo (f-i) = ','{:.4f}'.format(DthetaInf),' deg')
```
