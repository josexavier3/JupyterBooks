## Problema 1 <a name="PV121"></a>

O veio circular oco $AB$ representado na figura é feito de aço, o qual se assume ter comportamento elastoplástico perfeito com $\tau_C =$ 145 MPa e $G =$ 77 GPa. Aumenta-se gradualmente o valor $T$ do momento de torção até a zona plástica atingir a superfície interior do veio. Em seguida o momento de torção é removido. Represente a distribuição de tensões tangenciais residuais na secção transversal e determine o ângulo de torção permanente do veio.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i1.jpg"
width="500">

```python
import numpy as np

# Dados
# Unidades: m, N, Pa

L, ri, re = 1.25, 18.e-3, 30.e-3

TauC = 145.e6
G = 77.e9
```

### Resolução

Considere-se um veio circular de um material com comportamento elasto-plástico perfeito na relação tensão de corte - distorção angular ($\tau-\gamma$):

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i2.jpg"
width="400" height="260">

A partir desta lei de comportamento é possível determinar a função do momento de torção, $T$, em relação ao ângulo de torção, $\phi$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i3.jpg"
width="600" height="380">

Nesta lei identifique-se dois pontos:

- Ponto limite de cedência ($\phi_C,~T_C$)
- Ponto limite de plasticidade ($\phi,~T_P$)

**Ponto ($\phi_C,~T_C$):**

Desde que a tensão de corte não exceda a tensão de corte de cedência, a lei de Hooke é válida. De acordo com a fórmula de torção, a tensão de corte varia linearmente ao longo da secção, $\tau(\rho)$, e o momento máximo, $\tau_\mathrm{max}$, ocorre inicialmente na periferia do veio ($c = r_e$):

\begin{equation*}
\tau_\mathrm{max} = \frac{T c}{J}\quad\text{[MPa]}
\end{equation*}

O momento de torção que corresponde ao aparecimento da tensão de cedência (limite de elasticidade) na periferia do veio designa-se momento por torção máximo elástico, $T_C$. O seu valor pode ser calculado pela fórmula de torção considerando a tensão de corte de cedência $\tau_C$ ($c=r_e$):

\begin{equation*}
T_C = \frac{J \tau_C}{r_e}
\end{equation*}

em que $J$ é o momento polar de inércia de secção; para veios ocos:

\begin{equation*}
J = \frac{1}{2}\pi(r_e^4 -r_i^4)\quad\text{[m$^4$]}
\end{equation*}

```python
J = np.pi/2*(re**4 - ri**4)
print(f'J = {J:.3e} m3')
Tc = TauC*J/re
print(f'Tc = {Tc:.2f} N.m')
```

O ângulo de torção correspondente ao momento de torção crítico, $T_C$, será o ângulo de torção máximo elástico, $\phi_C$; este vem expresso por:

\begin{equation*}
\phi_C = \frac{T_C L}{JG}\quad\text{[radianos]}
\end{equation*}

```python
phiC = Tc*L/J/G
print(f'phiC = {phiC:.4f} radianos | {np.rad2deg(phiC):.2f} º')
```

De notar que o ângulo de torção crítico pode ainda ser obtido considerando a igualdade:

\begin{equation*}
\begin{aligned}
\phi_C r_e &= \gamma_C L ~\Leftrightarrow~\phi_C =  \frac{\tau_C}{G} \frac{L}{r_e}
\end{aligned}
\end{equation*}

```python
phiC2 = TauC/G*L/re
print(f'phiC (ver 2) = {phiC2:.4f} radianos | {np.rad2deg(phiC2):.2f} º')
```

**Ponto ($\phi,~T_P$):**

O progressivo incremento do momento torsor aplicado irá gerar uma deformação plástica. Tipicamente, a plastificação inicia no ponto na periferia do veio (onde a tensão de corte é superior) e migra paulatinamente (a tensão de corte máxima constante) para o seu interior. Desenvolve-se uma coroa circular de material plastificado, envolvendo um núcleo de material no domínio elástico. Designa-se por $\rho_C$ o raio do núcleo elástico. Na região plástica, a tensão é uniforme e iqual a $\tau_C$, enquanto que na região central elástica, a tensão de corte varia linearmente com $\rho$:

\begin{equation*}
\tau =
\begin{cases}
\displaystyle\frac{\rho}{\rho_C}\tau_C
~,&\quad
0 \le \rho \lt \rho_C ~\text{(domínio elástico)} \\
&\text{}\\
\tau_C~,&\quad
\rho_C \le \rho \le c ~\text{(domínio plástio)}
\end{cases}
\end{equation*}

Com o aumento do momento torsor $T$, eventualmente toda a região se expande para o domínio plástico à medida que $\rho_C \rightarrow 0$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i4.jpg"
width="900">

As tensões de corte existentes na secção devem equilibrar o momento de torção aplicado. Para obter esse momento torsor, $T$, devem integrar-se as tensões tangenciais ao longo da secção. Este cálculo pode ser feito utilizando coordenadas polares.

As expressões para a determinação do momento de torção em função da tensão de corte pode assim ser desenvolvido ($\,\rho_C \le \rho \le c ~-\text{ domínio plástio)}$:

- Veios maciços:

Para veios maciços, o momento máximo ou momento de torção plástico correspondente ao raio crítico, $\rho_C$ é dado pelo seguinte desenvolvimento:

\begin{equation*}
\begin{aligned}
T &= 2\pi \int_0^{\rho_C} \rho^2\left(\frac{\tau_C}{\rho_C}\rho\right)\mathrm{d}\rho +
2\pi\int_{\rho_C}^C \rho^2\tau_C\mathrm{d}\rho \\
&= \frac{2}{3}\pi c^3\tau_C\left(1-\frac{1}{4}\frac{\rho_C^3}{c^3}\right)
= \frac{4}{3}T_C\left(1-\frac{1}{4}\frac{\rho_C^3}{c^3}\right)
\end{aligned}
\end{equation*}

No limite quando a região plástica expande para a área de toda a secção, i.e., $\rho_C \rightarrow 0$:

\begin{equation*}
T_P = \frac{4}{3}T_C
\end{equation*}

- Veios ocos:

Pela análise do equilíbrio de momentos de torção resulta para veios ocos:

\begin{equation*}
\begin{aligned}
T = \int_A\rho\,\mathrm{d}F ~;~~ \mathrm{d}F = \tau\,\mathrm{d}A
\quad\Rightarrow\quad
T = \int_A\rho\,\tau\,\mathrm{d}A
\end{aligned}
\end{equation*}

Para um elemento de área definido em coordenadas polares ($\mathrm{d}A = \rho\mathrm{d}\theta\mathrm{d}\rho$):

\begin{equation*}
\begin{aligned}
T & = \int_A\rho\,\tau\,\mathrm{d}A = \int_{r_i}^{r_e}\int_0^{2\pi}\rho\,\tau\,(\rho\mathrm{d}\theta\mathrm{d}\rho) \\
& = \int_{r_i}^{r_e}\rho^2\,\tau[\theta]_0^{\,2\pi} \mathrm{d}\rho
= \left[\frac{\rho^3}{3}\right]_{r_i}^{r_e}\,\tau\,2\pi \mathrm{d}\rho
= \frac{2\pi}{3}(r_e^3-r_i^3)\tau\\
\end{aligned}
\end{equation*}

pelo que o momento torsor plástico (máximo que pode ser aplicado) para veios ocos é expresso por:

\begin{equation*}
T_P = \frac{2\pi}{3}\tau_C(r_e^3-r_i^3)
\end{equation*}

```python
Tp = 2*np.pi/3*TauC*(re**3-ri**3)
print(f'Tp = {Tp:.2f} N.m')
```

Como cada secção transversal do veio cilindro permanece plana e roda como um corpo rígido, é possível estabelecer uma relação entre a distorção angular, $\gamma$, com o ângulo de torção $\phi$ e a distância radial $\rho$, pelo comprimento de arco entre dois pontos periféricos do veio $AA'$, sendo:

\begin{equation*}
AA' = L\gamma = \rho\phi
~\Leftrightarrow~
\gamma = \frac{\rho\,\phi}{L}
\end{equation*}

Nesta expressão $\gamma$ e $\theta$ são medidos em radianos. Verifica-se assim que a distorção angular $\gamma$, é proporcional ao ângulo de torção $\phi$ e à distância radial $\rho$ a que um ponto material, num plano perpendicular ao eixo, se situa em relação ao eixo do veio. Por outro lado, conclui-se que, para um veio circular, a distorção angular varia linearmente com a distância ao eixo do veio.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i5.jpg"
width="650" height="480">

A distância radial crítica, $\rho_C$ (raio que define a extensão da secção central onde permanece válido o regime linear elástico) pode ser obtida considerando a distorção angular crítica $\gamma_C$:

\begin{equation*}
\rho_C = \frac{L\,\gamma_C}{\phi}
\end{equation*}

Usando o ângulo de torção no início da cedência $\phi_C$ (isto é, quando $\rho_C = c$, para um ponto na periferia do veio) e fazendo $\phi = \phi_C$ quando $\rho_C = c$, a expressão anterior toma a forma:

\begin{equation*}
c = \frac{L\,\gamma_C}{\phi_C}
\end{equation*}

ou seja, em resumo:

\begin{equation*}
\begin{cases}
\rho_C\, \phi = L\,\gamma_C, & \quad \rho = \rho_C \\
c\, \phi_C = L\,\gamma_C, & \quad \rho = c \\
\end{cases}
\end{equation*}

Dividindo as últimas expressões, termo a termo, resulta a seguinte relação:

\begin{equation*}
\frac{\rho_C}{c} = \frac{\phi_C}{\phi}
\end{equation*}

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i6.jpg"
width="600" height="300">

Esta última expressão pode ser resolvido para calcular o ângulo de torção total sofrido pelo veio pela aplicação do momento de torção $T$:

\begin{equation*}
\phi = \left(\frac{c}{\rho_C}\right) \phi_C
\end{equation*}

Para o caso de um veio oco a expressão anterior escreve-se:

\begin{equation*}
\phi = \left(\frac{r_e}{r_i}\right) \phi_C
\end{equation*}

```python
phi = re/ri*phiC
print(f'phi = {phi:.4f} radianos | {np.rad2deg(phi):.2f} º')
```

Se o momento torsor for removido, após aplicação de um momento com valor superior ao momento elástico máximo, a redução de tensão de corte acontecerá ao longo de um segmento de reta na curva tensão-deformação de corte:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i9.jpg"
width="600">

O valor final da tensão de corte não será em geral igual a zero, devido a tensões residuais entretanto instaladas.

No diagrama momento torsor - distorção angular ($T-\phi$) a parte de descarga do momento aplicado será representado por um segmento de reta, com declive igual ao troço linear. De notar que o ângulo de torção não volta a zero depois da remoção do momento torsor aplicado, resultando uma deformação angular permanente obtida por:

\begin{equation*}
\begin{aligned}
\phi' = \phi - \phi_R = \frac{T_p L }{JG}
\end{aligned}
\end{equation*}

em que,

- $\phi~:~$é o ângulo de torção associado ao momento torsor máximo aplicado (neste caso o momento plástico);
- $\phi'~:~$é o ângulo de torção na descarga assumindo um comportamento linear elástico (recuperação elástica): $T_pL/JG$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P001_i10.jpg"
width="675" height="380">

```python
phidescarga  = Tp*L/J/G
print(f'phi(descarga) = {phidescarga:.4f} radianos | {np.rad2deg(phidescarga):.2f} º')
```

Eventualmente o valor de $\phi'$ pode ainda ser obtido por (relação geométrica de igualdade de triângulos):

\begin{equation*}
\begin{aligned}
\frac{T_C}{\phi_C} = \frac{T_P}{\phi\,{'}}
\end{aligned}
\end{equation*}

```python
phidescarga2  = Tp/Tc*phiC
print(f'phi(descarga) (vers.2) = {phidescarga2:.4f} radianos | {np.rad2deg(phidescarga2):.2f} º')
```

Pelo que a tensão residual pode ser obtida por:

\begin{equation*}
\begin{aligned}
\phi_R  = \phi - \phi'
\end{aligned}
\end{equation*}

```python
phiresidual = phi - phidescarga
print(f'phiresidual = {phiresidual:.4f} radianos | {np.rad2deg(phiresidual):.2f} º')
```

**Princípio de superposição**

As tensões residuais no material elastoplástico são obtidas por aplicando o princípio de superposição.

Por um lado, consideram-se as tensões devidas à aplicação do momento torsor, $T$ (envolvendo plastificação). Por outro lado, são consideradas as tensões devido a um momento de torção, igual e oposto, que é aplicado na fase de descarga assumindo comportamento linear elástico.

O primeiro grupo de tensões reflete o comportamento elasto-plástico do material durante a fase de carregamento. O segundo grupo tem o comportamento linear do mesmo material durante o estágio de descarga. A adição dos dois grupos de tensões fornece a distribuição das tensões residuais no eixo.

Na descarga, a tensão mínima ($\tau_\mathrm{min}$) e máxima ($\tau_\mathrm{max}$) lineares no interior ($r_i$) e exterior ($r_e$) do tubo vêm:

\begin{equation*}
\begin{cases}
\tau_\mathrm{max} &= \displaystyle\frac{(-T_P) r_e}{J} \\
\tau_\mathrm{min} &= \displaystyle\frac{(-T_P) r_i}{J} =
\tau_\mathrm{max} \displaystyle\frac{r_i}{r_e}\\
\end{cases}
\end{equation*}

```python
taumaxDescarga =  -Tp*re/J
print(f'tau max (Descarga) = {taumaxDescarga*1e-6:.2f} MPa')

tauminDescarga =  -Tp*ri/J
print(f'tau min (Descarga) = {tauminDescarga*1e-6:.2f} MPa')

tauminDescarga2 =  taumaxDescarga*ri/re
print(f'tau min (Descarga) (versão 2) = {tauminDescarga2*1e-6:.2f} MPa')
```

Assim sendo as tensões residuais têm uma distribuição linear cujos valores no ponto interior e exterior ao veio oco são determinados por:

\begin{equation*}
\begin{cases}
\tau_i^R = \tau_C + \tau_\mathrm{min} = \tau_C - \displaystyle\frac{T_P
r_i}{J} \\
\tau_e^R = \tau_C + \tau_\mathrm{max} = \tau_C - \displaystyle\frac{T_P
r_e}{J}
\end{cases}
\end{equation*}

```python
tauresidualri = TauC + tauminDescarga
print(f'tau Residual (ri) = {tauresidualri*1e-6:.2f} MPa')

tauresidualre = TauC + taumaxDescarga
print(f'tau Residual (re) = {tauresidualre*1e-6:.2f} MPa')
```

De notar que as tensões residuais são positivas ou negativas em zonas distintas na secção do veio. Esta condição é necessária para se verificar a condição de que:

\begin{equation*}
T = \int \rho(\tau\mathrm{d}A) = 0
\end{equation*}
