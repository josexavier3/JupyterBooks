## Problema 5 <a name="PIV95"></a>

Um veio de aço de 50 mm de diâmetro e um tubo de alumínio de 74 mm de diâmetro exterior e 8 mm de espessura estão ligados a um apoio fixo e a um disco rígido, como se mostra na figura. Sabendo que as tensões iniciais são nulas e que a resistência enrolada no veio de aço faz com que a sua temperatura se eleve em 280º C, determine:

[a)](#A) As tensões no veio e no tubo.

[b)](#B) O alongamento final do conjunto.

Dados: $E_\mathrm{aço} = $ 200 GPa,
$E_\mathrm{AI} = $ 70 GPa, $\alpha =
15\times10^{-6}$ ºC$^{-1}$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au09/Au09_P005_i1.jpg"
width="550">


## Resolução

#### a) As tensões no veio e no tubo.<a name="A"></a> ([_enunciatum_](#PIV95))

```python
# units: mm, MPa,N, ºC

import numpy as np

# dados: veio
dv, Ev = 50., 200.e3

## dados: tubo
dtext = 74.
et = 8.
dtint = dtext - 2*et
print(f'd int tubo : {dtint:6.1f} mm')
Et = 70.e3

L = 500.
coefDilata = 15.e-6
DeltaT = 280.

Av = np.pi*dv**2/4
print(f'Area veio : {Av:6.1f} mm^2')

At = np.pi/4*(dtext**2 - dtint**2)
print(f'Area veio : {At:6.1f} mm^2')
```

**Configuração deformada**

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au09/Au09_P005_i2.jpg"
width="750">

**Equação de compatibilidade**

\begin{equation*}
\delta_v^T - \delta_v^F = \delta_t^F
\end{equation*}

**Equilíbrio estático**

\begin{equation*}
\sum F = 0
\quad\Leftrightarrow\quad
-F' + F'' = 0
\quad\Leftrightarrow\quad
F' = F'' = F
\end{equation*}

**Sistema de equações final**

\begin{equation*}
\alpha \Delta T L - \frac{FL}{A_v E_v} = \frac{FL}{A_t E_t}
\quad\Leftrightarrow\quad
F = \frac{\alpha \Delta T}{\left(\displaystyle\frac{1}{A_t E_t}+\displaystyle\frac{1}{A_v
E_v}\right)}
\end{equation*}

```python
F = coefDilata*DeltaT/(1/Av/Ev + 1/At/Et)
print(f'F = {F:6.2f} N')

Sv = F/Av
print(f'Sv = {Sv:6.2f} MPa')
St = F/At
print(f'St = {St:6.2f} MPa')
```

#### b) O alongamento final do conjunto.<a name="B"></a> ([_enunciatum_](#PIV95))

```python
deslot = F*L/At/Et
print(f'Deslocamento final tubo = {deslot:6.2f} mm')
deslov = coefDilata*DeltaT*L - F*L/At/Et
print(f'Deslocamento final veio = {deslot:6.2f} mm')
```
