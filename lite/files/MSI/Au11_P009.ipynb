## Problema 9 <a name="PV119"></a>
***

Três veios de secção circular uniforme, representados na figura, com um diâmetro $\phi = 18$ mm, são construídos em aço, com $G = 77$ GPa. Sabendo que as duas extremidades A e D estão fixas, determine a intensidade das tensões tangenciais máximas que ocorrerão em cada um dos veios quando o momento ME provocar na extremidade E do veio EF uma rotação de 1$^\circ$.

## Resolução


```python
import numpy as np
from typing import NamedTuple

# Dados
# unidades: m, Pa, N
class dados(NamedTuple):
    diam : float # m
    G : float # Pa
    rotEF : float  # deg
    rB: float # m
    rC: float # m
    rF: float # m
    lAB: float# m
    lCD: float# m

d = dados(18.e-3, 77.e9, np.deg2rad(1.), 30.e-3, 120.e-3, 40.e-3, 1.6, 1.2)

def Jp(r):
    return np.pi*(r**4)/2

J = Jp(d.diam/2)
print(f'{J:.4e} mm4')
```

**Eq. Estático: Diagram de corp livre dos veios**

#### Troço AB:

\begin{equation*}
\sum M_t = 0
\quad\Leftrightarrow\quad
M_A = r_B F^{''}
\end{equation*}

#### Troço CD:

\begin{equation*}
\sum M_t = 0
\quad\Leftrightarrow\quad
M_D = r_C F^{'} - r_C F^{''}
\end{equation*}

#### Troço EF:

\begin{equation*}
\sum M_t = 0
\quad\Leftrightarrow\quad
M_E = r_F F^{'}
\end{equation*}

Combinando,

\begin{equation*}
M_D = \frac{r_C}{r_F} M_E  -  \frac{r_C}{r_B} M_A
\end{equation*}

ou,

\begin{equation*}
M_E = \frac{r_F}{r_C} M_D  +  \frac{r_F}{r_B} M_A
\end{equation*}

**Cinemática**

\begin{equation*}
\phi_B = \frac{M_Al_{AB}}{JG}
\quad\wedge\quad
\phi_C = \frac{M_Dl_{CD}}{JG}
\end{equation*}

\begin{equation*}
\phi_B = \frac{r_C}{r_B}
\phi_C
\end{equation*}

Resultando,

\begin{equation*}
M_A = \left(\frac{r_C}{r_B}\right)
\left(\frac{l_{CD}}{l_{AB}}\right)
M_D
\end{equation*}

Por outro lado,

\begin{equation*}
M_E = \left( \frac{r_F}{r_C}  + \frac{r_F}{r_B}
\frac{r_C}{r_B} \frac{l_{CD}}{l_{AB}} \right) M_D
\end{equation*}

\begin{equation*}
\phi_E = \phi_{E/F} + \frac{r_C}{r_F}\phi_{C/D}
\end{equation*}


\begin{equation*}
\phi_{E/F} = \frac{M_EL_{AB}}{JG}
\quad\wedge\quad
\phi_{C/D} = \frac{M_DL_{CD}}{JG}
\end{equation*}

logo,

\begin{equation*}
\phi_E = \frac{M_E L_{EF}}{JG} + \left( \frac{r_C}{r_F}\right) \frac{M_D L_{CD}}{JG}  = 1^\circ
\end{equation*}

```python
import sympy as sp

ma, me, md = sp.symbols('ma me md')

eq1 = md - d.rC/d.rF*me + d.rC/d.rB*ma
eq2 = me - d.rF/d.rC*md - d.rF/d.rB*ma
eq2_ = ma - (d.rC/d.rB)*(d.lCD/d.lAB)*md
eq3 = ma - d.rC/d.rB*d.lCD/d.lAB*md
eq3_ = me - (d.rF/d.rC + d.rF/d.rB*d.rC/d.rB*d.lCD/d.lAB)*md
eq4 = d.rotEF - me*d.lAB/d.G/J - d.rC/d.rF*md*d.lCD/d.G/J

sol1 = sp.solve({eq3_,eq4},{me,md})
ME = sol1[me]
print(f'ME = {ME:.3f} N.m')
MD = sol1[md]
print(f'MD = {MD:.3f} N.m')

sol2 = sp.solve([eq2.subs({me: ME, md: MD}), eq4], [ma])
MA = sol2[ma]
print(f'MA = {MA:.3f} N.m')
```

```python
def tauval(m,r,J):
    return m*r/J

tauME = tauval(ME,d.diam/2,J)
print(f'tauME = {tauME*1e-6:.3f} MPa')
tauMD = tauval(MD,d.diam/2,J)
print(f'tauMD = {tauMD*1e-6:.3f} MPa')
tauMA = tauval(MA,d.diam/2,J)
print(f'tauMA = {tauMA*1e-6:.3f} MPa')
```
