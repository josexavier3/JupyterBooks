## Problema 6 <a name="PV136"></a>

Os veios AB e CD representados na figura são ambos construídos em aço, que é considerado um material elastoplástico com $G = 77$ GPa e $\tau_c = 168$ MPa, têm um comprimento respectivamente igual a 300 mm e a 500 mm e um raio de 20,6 mm. Devido a uma avaria ocorrida no motor que está ligado à extremidade D, o momento de torção $T = 1000$ N.m é aplicado nessa extremidade, enquanto a extremidade A está fixa.

[a)](#A) Determine a distribuição de tensões tangenciais nos dois veios, representando graficamente essa distribuição e calculando a tensão tangencial máxima em cada veio.

[b)](#B) Calcule a rotação que ocorre na secção D quando o momento T é aplicado.

[c)](#C) Diga o que entende por tensões residuais e por deformações residuais.

[d)](#D) Considere agora que o momento T é retirado, e determine a distribuição de tensões residuais nos dois veios, assim como a rotação residual da extremidade D.

```python
import numpy as np
from typing import NamedTuple

# Dados
# unidades: m, Pa, N
class dados(NamedTuple):
    G : float # kN
    tauC : float # MPa
    lAB : float  # m
    lCD : float
    c : float
    T : float # N.m
    rC : float # N.m
    rB : float # N.m

d = dados(77.e9, 168.e6, 0.3, 0.5, 20.6e-3, 1.e3, 40.e3, 100.e3)
```

## Resoluçaão

### a) Determine a distribuição de tensões tangenciais nos dois veios, representando graficamente essa distribuição e calculando a tensão tangencial máxima em cada veio.


**Diagrama de corpo livre**


**Veio AB + Engrenagem**

\begin{equation*}
\sum M = 0
\quad\Leftrightarrow\quad
T_A = r_A F
\quad\Leftrightarrow\quad
F = \frac{T_A}{r_A}
\end{equation*}

**Veio CD + Engrenagem**

\begin{equation*}
\sum M = 0
\quad\Leftrightarrow\quad
T = r_C F
\quad\Leftrightarrow\quad
F = \frac{T}{r_C}
\end{equation*}

Desenvolvendo,

\begin{equation*}
\frac{T_A}{r_A} = \frac{T}{r_C}
\quad\Leftrightarrow\quad
T_A = \left(\frac{r_B}{r_A}\right) T
\end{equation*}


```python
TA = (d.rB/d.rC)*d.T
print(f"TA = {TA:.1f} N.m")
```

**Cinemática da engregangem**

\begin{equation*}
\phi_B r_B = \phi_C r_C
\quad\Leftrightarrow\quad
\phi_C = \frac{r_B}{r_C} \phi_B
\end{equation*}

**Tensão de corte máxima**

\begin{equation*}
\tau_{AB,min} = \frac{Tc}{J}
\end{equation*}


```python
def Jpolar(r): # momento de inércia de área polar
    return np.pi*r**4/2

J = Jpolar(d.c)
print(f"J = {J:.3e} m4")

tauABmax = TA*d.c/J
print(f"tauABmax = {tauABmax*1e-6:.1f} MPa > {d.tauC*1e-6:.1f} MPa")
```

Uma vez que $\tau_{AB,min} > \tau_\textrm{ced}$, conclui-se que o veio AB está no domínio plástico.

\begin{equation*}
T_{AB} = \frac{4}{3} T_C \left[1-\frac{1}{4}\left(\frac{\rho_C}{c}\right)³ \right]
\end{equation*}

em que,

Tensão de corte cedência:

\begin{equation*}
T_{C} = \frac{J}{c} \tau_C
\end{equation*}

Tensão de corte plástica:

\begin{equation*}
T_p = \frac{4}{3}T_C
\end{equation*}




```python
TC = J/d.c*d.tauC
print(f"TC = {TC:.1f} N.m")

Tp = 4/3*TC
print(f"Tp = {Tp:.1f} N.m")
```

\begin{equation*}
\rho_C³= 4 c³ \left[ 1-\frac{3}{4} \left( \frac{T_{AB}}{T_C} \right) \right]
\end{equation*}

```python
rhoC = (4*d.c**3*(1-3/4*(TA/TC)))**(1/3)
print(f"rhoC = {rhoC*1e3:.1f} mm")
```

**Tensão de corte máxima**
**Veio CD**

\begin{equation*}
\tau_{CD,min} = \frac{T c}{J}
\end{equation*}

```python
tauCDmin = d.T*d.c/J
print(f"tauCDmin = {tauCDmin*1e-6:.1f} MPa < {d.tauC*1e-6:.1f} MPa")
```

Como $\tau_{CD,min}$ é menor do que a tensão de corte de cedência o comportamento é elástico.

### b) Calcule a rotação que ocorre na secção D quando o momento T é aplicado.

\begin{equation*}
\phi_{D/C} = \phi_{D} - \phi_{C} = \frac{T_{CD}L_{CD}}{JG}
\quad\wedge\quad
\phi_{C} = \frac{r_B}{r_C} \phi_{B}
\end{equation*}

**Veio AB (no domínio plástico)**

\begin{equation*}
\frac{\rho_C}{c} = \frac{\phi_C}{\phi}
\quad\wedge\quad
\phi_C = \frac{T_C L_{AB}}{JG}
\end{equation*}

resolvendo,

\begin{equation*}
\phi = \frac{c}{\rho_C} \phi_C}
\end{equation*}


```python
phiC = TC*d.lAB/J/d.G
print(f"phi_C = {phiC:.4f} rad | {phiC*180/np.pi:.1f} deg")
phi_B = d.c/rhoC*phiC
print(f"phi_B = {phi_B:.4f} rad | {phi_B*180/np.pi:.1f} deg")

phi_C = d.rB/d.rC*phi_B
print(f'phi_C = {phi_C:.4f} rad | {phi_C*180/np.pi:.2f} deg')

phi_D = phi_C + d.T*d.lCD/J/d.G
print(f'phi_D = {phi_D:.4f} rad | {phi_D*180/np.pi:.2f} deg')
```

### d) Considere agora que o momento T é retirado, e determine a distribuição de tensões residuais nos dois veios, assim como a rotação residual da extremidade D.

**Veio AB**

\begin{equation*}
\phi_{AB} = \frac{T_{AB}L_{AB}}{JG}
\quad\wedge\quad
\frac{\rho_{AB}^c}{c_{AB}} = \frac{\phi_{AB}^c}{\phi_{AB}}
\end{equation*}

\begin{equation*}
\phi_{AB}^{'} = \phi_{AB} - \phi_{AB}^{R}=\frac{T_{AB} L_{AB}}{JG}
\end{equation*}


```python
phiAB_ = TA*d.lAB/J/d.G
print(f"phiAB_ = {phiAB_:.4f} rad | {phiAB_*180/np.pi:.2f} deg")
```

```python
phiAB_R = phi_B - phiAB_
print(f"phiAB_R = {phiAB_R:.3e} rad | {phiAB_R*180/np.pi:.3f} deg")
```

**Veio CD**

\begin{equation*}
\phi_{C}^R = \frac{r_B}{r_C} \phi_{B}^R
\end{equation*}

```python
phiC_R = d.rB/d.rC*phiAB_R
print(f"phiC_R = {phiC_R:.3e} rad | {phiC_R*180/np.pi:.2f} deg")
```

**Tensão de corte residual**

\begin{equation*}
\tau_1 = \tau_\textrm{ced} - \tau_\textrm{max}
\quad\wedge\quad
\tau_2 = \tau_\textrm{ced} - \tau(\rho_C)
\quad\wedge\quad
\tau(\rho = \rho_C) \quad\leftrightarrow\quad
\tau(\rho_C) = \frac{\rho_C}{c}\tau_\textrm{max}
\end{equation*}

```python
tau1 = d.tauC - tauABmax
print(f"tau1 = {tau1*1e-6:.2f} MPa")
taurhoC = rhoC/d.c*tauABmax
print(f"taurhoC = {taurhoC*1e-6:.2f} MPa")
tau2 = d.tauC - taurhoC
print(f"tau2 = {tau2*1e-6:.2f} MPa")
```
