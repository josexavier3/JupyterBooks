## Problema 6 <a name="PV126"></a>

Um aparelho $F$, destinado a registar em formato digital a rotação do veio $A$, está ligado a esse veio através do conjunto de engrenagens indicado na figura, que consiste em quatro rodas dentadas e três veios maciços de aço, cada um com um diâmetro $d$. Duas das rodas têm um raio $r$ e as outras duas um raio $nr$. Se a rotação do aparelho $F$ for impedida, determine o ângulo de rotação da extremidade $A$ em função de: $T_A, l, G, J, n$.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au11/Au11_P006_i1.jpg"
width="400" height="270">


## Resolução

**Cinemática das engrenagens**

\begin{equation*}
\begin{aligned}%
r_E\,\phi_E &= r_D\,\phi_D
\quad (r_E \equiv r~;~r_D \equiv nr)\quad \\% 
\quad\Leftrightarrow\quad r\,\phi_E &= n r \,\phi_D \\%
\quad\Leftrightarrow\quad  \phi_E &= \frac{r_D}{r_E} \phi_D = n\,\phi_D
\quad\wedge\quad
r_C\,\phi_C = r_B\,\phi_B
\end{aligned}%
\end{equation*}

\begin{equation*}
\quad\Leftrightarrow\quad
\phi_C = n\,\phi_B
\quad\wedge
\quad r_D = r_B \quad\wedge\quad r_C = r_E 
\end{equation*}


**Relação de momentos de torção**

\begin{equation*}
T_{AB} = T_A
\quad\wedge\quad
T_{CD} =\frac{r_C}{r_B} T_{AB} = \frac{T_{AB}}{n} = \frac{T_A}{n}
\quad\wedge\quad
T_{EF} =\frac{r_E}{r_D} T_{CD} = \frac{T_{CD}}{n} = \frac{T_A}{n^2}
\end{equation*}

**Relação de ângulos de torção**

\begin{equation*}
\begin{aligned}
\phi_{E} = \phi_{EF} = \frac{T_{EF}L_{EF}}{GJ} = \frac{T_{A}L}{n^2GJ} \\%
\quad\wedge\quad 
\phi_{D} = \frac{r_E}{r_D} \phi_{E} = \frac{\phi_{E}}{n} =
\frac{T_{A}L}{n^3GJ} \\%
\quad\wedge\quad 
\phi_{CD} = \frac{T_{CD}L_{CD}}{GJ} = \frac{T_{A}L}{nGJ}
\end{aligned}
\end{equation*}


\begin{equation*}
\phi_{C} = \phi_{D} + \phi_{CD} = \frac{T_{A}L_{EF}}{n^3GJ}
+ \frac{T_{A}L_{EF}}{nGJ}
= \frac{T_{A}L}{GJ}~\left(\frac{1}{n^3}+\frac{1}{n}\right)
\end{equation*}

\begin{equation*}
\phi_{B} = \frac{r_C}{r_B} \phi_{C} = \frac{\phi_{C}}{n}
= \frac{TL}{GJ}~\left(\frac{1}{n^4}+\frac{1}{n^2}\right)
\end{equation*}

\begin{equation*}
\phi_{AB} = \frac{T_{AB}L_{AB}}{GJ} = \frac{T_{A}L}{GJ}
\end{equation*}

\begin{equation*}
\phi_{A} = \phi_{B} + \phi_{AB}
= \frac{T_{A}L}{GJ} \left(\frac{1}{n^4}+\frac{1}{n^2}+1\right)
\end{equation*}
