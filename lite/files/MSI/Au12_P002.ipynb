## Problema 2 <a name="PV122"></a>

O veio circular maciço indicado na figura é feito de aço, o qual se assume ter comportamento elastoplástico perfeito, com $\tau_C = 145$ MPa e $G = 77$ GPa. O momento torsor $T$ é aumentado gradualmente até o ângulo de torção do veio atingir 6 $^\circ$, sendo removido em seguida. Determine:

[a)](#A) O valor e a localização da tensão tangencial residual máxima.

[b)](#B) O ângulo de torção permanente do veio.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P002_i1.jpg"
width="400" height="250">



```python
from typing import NamedTuple
import numpy as np

# Dados
# unidades: m, Pa, N
class dados(NamedTuple):
    tauC : float # kN
    G : float # kN.m
    rot : float  # deg
    L : float  # m
    r : float  # m

d = dados(145.e6, 77.e9, np.deg2rad(6), 0.6, 16.e-3)

```

### a) O valor e a localização da tensão tangencial residual máxima.


```python
gammaC = d.tauC/d.G
print(f'{gammaC:.4e} rad')
```

**Raio do núcleo elástico**

\begin{equation*}
\rho_C = \frac{L\gamma_C}{\phi}
\end{equation*}

```python
rhoe = d.L*gammaC/d.rot
print(f'{rhoe:.4e} m')
```

**Tensão de corte de cedência**

\begin{equations*}
T_C = \frac{J}{c}\tau_C
\end{equations*}

```python
J = np.pi*d.r**4/2
print(f'J = {J:.4e} m4')

TC = J/d.r*d.tauC
print(f'TC = {TC:.2f} N.m')
```

#### b) O ângulo de torção permanente do veio.

```python
T = 4/3*TC*(1-1/4*(rhoe/d.r)**3)
print(f'T = {T:.2f} N.m')
```

```python
Tp = 4/3*TC
print(f'Tp = {Tp:.2f} N.m')
```

```python
phi_ = T*d.L/J/d.G
print(f'phi = {phi_:.4e} rad | {np.rad2deg(phi_):.3f} deg')
```

```python
phi_r = d.rot - phi_
print(f'phi_residual = {phi_r:.4e} rad | {np.rad2deg(phi_r):.3f} deg')
```

\begin{equations*}
\tau^{'} = \frac{T}{J}\rho
\end{equations*}

```python
def taur(r):
    return T/J*r

tau1 = taur(rhoe)
print(f' tau1 = {tau1*1e-6:.2f} MPa')

tau2 = taur(d.r)
print(f' tau2 = {tau2*1e-6:.2f} MPa')
```

```python
tauR1 = d.tauC - tau1
print(f' tauR1 = {tauR1*1e-6:.2f} MPa')

tauR2 = d.tauC - tau2
print(f' tauR2 = {tauR2*1e-6:.2f} MPa')
```
