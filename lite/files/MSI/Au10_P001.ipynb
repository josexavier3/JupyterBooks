### a) Diga o que entende por deformações e por tensões residuais. <a id="A"></a> ([_enunciatum_](#PIV101))

- Deformações residuais são deformações elásticas introduzidas no material por tensões residuais, e que deformam permanentemente a configuração nominal ou de referência  de um objeto.

- As tensões residuais são tensões que permanecem na estrutura em equilíbrio depois do carregamento (mecânico ou térmico) ser removido. Em alguns casos, as tensões residuais geram um deformação plástica permanente, levando ao empeno e distorção de um objeto (desvios em relação à configuração nominal). Noutros casos, as tensões residuais afetam a integridade estrutural das peças pela suscetibilidade de ocorrência de rutura por fratura e fadiga.

- As tensões residuais consistem nas tensões que existem no interior de um corpo sólido elástico na ausência de, ou em adição a, tensões devidas a esforços aplicados exteriormente. Podem ser à escala microscópica ou à escala macroscópica. As do tipo macroscópico são geralmente provocadas por desajustamento estrutural entre componentes de uma mesma estrutura. À escala microscópica, são originadas por deformações desiguais dos cristais ou grãos do material (tensões térmicas) durante operações de aquecimento ou arrefecimento das peças (por exemplo, soldadura, têmpera, etc.) ou devido a processos de deformação plástica não homogénea realizados a frio (por exemplo, estampagem, maquinagem, etc.). Podem ser medidas usando a difração de raios X (métodos não destrutivos) ou recorrendo a extensómetros elétricos através da técnica de relaxação de tensões por seccionamento, por furacão ou por desbaste (métodos destrutivos).

### b) Determine a tensão normal em cada tirante quando a força $\vec{P}$ é aplicada na extremidade C. <a id="B"></a> ([_enunciatum_](#PIV101))

**Equilíbrio estático**

Considere-se o diagrama de corpo livre (DCL) da barra rigida:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au10/Au10_P001_i2.jpg"
width="600">

A análise do equilíbrio estático tendo por base as forças representadas no DCL da barra rigida ABC permite escrever:

\begin{equation*}
\begin{cases}
\uparrow(+)&~\sum \vec{F}_y=0~:& \vec{F}_\textrm{BD} - \vec{F}_\textrm{AE} -
\vec{P} = 0
\\
\circlearrowleft(+)&\sum M_C=0:& (-l_\textrm{CA} \vec{\imath})\times
(-F_\textrm{AE}\,
\vec{\jmath}) +
(-l_\textrm{CB} \vec{\imath})\times(F_\textrm{BD}\,\vec{\jmath}) =
0\vec{k}
\end{cases}
\end{equation*}

\begin{equation*}
\Leftrightarrow
\begin{cases}
F_\textrm{BD} = F_\textrm{AE} + P \\
 l_\textrm{CA}\,F_\textrm{AE}
- l_\textrm{CB}\,F_\textrm{BD} = 0
\end{cases}
\Leftrightarrow
\begin{cases}
F_\textrm{BD} = F_\textrm{AE} + P \\
 \,F_\textrm{AE} = \displaystyle\left(\frac{l_\textrm{CB}}{l_\textrm{CA}}\right)
\,F_\textrm{BD}
\end{cases}
\end{equation*}

Neste problema existem mais incógnitas do que equações de equilíbrio, pelo que é um problema dito estaticamente indeterminado. Para a solução do problema outras equações devem ser definidas.

**Equações de compatibilidade dos deslocamentos**

Considere-se a deformada da barra rígida, assumindo um comportamento global linear elástico, de acordo com o esquema seguinte:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au10/Au10_P001_i3.jpg"
width="500" height="295">

Partindo da hipótese da geometria deformada da barra rigida ABC, é possível definir as seguintes equações de compatibilidade de deslocamentos adicionais (por igualdade entre triângulos retos):

\begin{equation*}
\tan\theta = \frac{\delta_\textrm{A}}{x}
= \frac{\delta_\textrm{B}}{l_\textrm{AB} - x}
= \frac{\delta_\textrm{C}}{l_\textrm{AC} - x}
\end{equation*}

**Relação força-deslocamento (axial) para barras elásticas**

Considerando agora as relações força-deslocamento axiais para a barra elástica é possível concluir:

\begin{equation*}
\frac{\delta_\textrm{A}}{\delta_\textrm{B}}
= \frac{x}{l_\textrm{AB} - x}
\quad\wedge\quad \delta_\textrm{A} = \frac{F_\textrm{AE} l_\textrm{AE}}{AE}
\quad\wedge\quad \delta_\textrm{B} = \frac{F_\textrm{BD} l_\textrm{BD}}{AE}
\end{equation*}

Desenvolvendo,

\begin{equation*}
\begin{aligned}
&\quad\frac{F_\textrm{AE} l_\textrm{AE}}{F_\textrm{BD} l_\textrm{BD}} =
\frac{x}{l_\textrm{AB} - x}
\quad\wedge\quad
\frac{F_\textrm{AE}}{F_\textrm{BD}} = \displaystyle\left
(\frac{l_\textrm{CB}}{l_\textrm{CA}}\right)\\
\Rightarrow
&\quad\displaystyle\left
(\frac{l_\textrm{CB}}{l_\textrm{CA}}\right)
\frac{l_\textrm{AE}}{l_\textrm{BD}} =
\frac{x}{l_\textrm{AB} - x}
\end{aligned}
\end{equation*}

Esta última equação pode ser resolvido para $x$,

\begin{equation*}
x = \frac{l_\textrm{CB}l_\textrm{AE}l_\textrm{AB}}{l_\textrm{CA}l_\textrm{BD} + l_\textrm{CB}l_\textrm{AE}}
\end{equation*}

```python
# untis: mm, MPa, N, ºC

import sympy as sy
import numpy as np

# dados
lCB, lAB, lAE, lBD = 350., 250., 200., 300.
lCA = lAB + lCB
E, Sced, A = 210.e3, 250., 6.*30.
print(f'A = {A:3.1f} mm2')
AE = A*E
print(f'AE = {AE*1e-3:3.1f} kN')

dC = 1.5

X2 = (lCB*lAE*lAB)/(lCA*lBD+lCB*lAE)
print(f' x (versão 1) = {X2:3.1f} mm')

x = sy.symbols('x')
EQ = (lCB*lAE)/(lCA*lBD) - x/(lAB-x)
sol = sy.solve(EQ,{x})
X = sol[0]
print(f' x (versão 2) = {X:3.1f} mm')
```

Assim sendo, os deslocamentos nos pontos A e B são calculados por:

\begin{equation*}
\delta_\textrm{A}
= \frac{x\,\delta_\textrm{C}}{l_\textrm{AC} - x}
\quad\wedge\quad
\delta_\textrm{B}
= \frac{l_\textrm{AB} - x}{l_\textrm{AC} - x} \delta_\textrm{C}
\end{equation*}

```python
dA = X*dC/(lCA - X)
print(f' dA = {dA:6.3f} mm')

dB = (lAB-X)/(lCA - X)*dC
print(f' dB = {dB:6.3f} mm')
```

Conhecido o incremento de deslocamento axial a que uma barra é submetida (bem como geometria da secção, comprimento e propriedades elásticas da barra) é possível determinar a força axial geradora dessa deformação:

\begin{equation*}
F_\textrm{AE}  = \frac{\delta_\textrm{A} AE}{l_\textrm{AE}}
\quad\wedge\quad F_\textrm{BD} = \frac{\delta_\textrm{B} AE}{l_\textrm{BD}}
\quad\wedge\quad P = F_\textrm{BD} - F_\textrm{AE}
\end{equation*}

```python
FAE = dA*AE/lAE
print(f'FAE = {FAE:6.2f} N')

FBD = dB*AE/lBD
print(f'FBD = {FBD:6.2f} N')

P = FBD - FAE
print(f'P = {P:6.2f} N')
```

**Força admissível**

Atendendo à força máxima admissível nas barras AE e BD (construidas com o mesmo material) pela lei de comportamento elasto-plástico perfeito têm-se que:

\begin{equation*}
F_\textrm{max} = \sigma_\textrm{ced} A
\end{equation*}

```python
Fmax = Sced*A
print(f'Fmax = {Fmax:6.1f} N')
```

É possível portanto concluir que, para o deslocamento em C imposto (ou para a força $P$ aplicada para esse deslocamento), a barra BD entra em cedência pois $F_\textrm{BD} > F_\textrm{max}$.

De acordo com a lei de comportamento elasto-plástico perfeito, por hipótese, a barra BD ao longo de todo o domínio plástico está submetida a uma força constante máxima: $F_\textrm{BD} = F_\textrm{max} = 45$ kN.

De notar que a relação: $\delta_\textrm{B} = F_\textrm{BD} l_\textrm{BD}/AE$ deixa de ser válida para a barra BD, a partir do instante de cedência, e por isso não pode ser usada para a determinação do deslocamento total do ponto B, $\delta_\textrm{B,tot}$. Contudo, destaca-se que as relações geométricas da deformada da estrutura permanecem válidas conduzindo às mesmas expressões gerais de compatibilidade de deslocamentos.

**Plastificação da barra BD**

A análise anterior deve ser portanto atualizada, considerando a plastificação da barra BD.

Do equilíbrio estático (dos momentos das forças aplicadas em relação a C) verifica-se, como se viu anteriormente, a relação:

\begin{equation*}
F_\textrm{AE} = \displaystyle\left(\frac{l_\textrm{CB}}{l_\textrm{CA}}\right)
\,F_\textrm{BD} \Bigg|_\textrm{plastificação da barra BD} =
\displaystyle\left
(\frac{l_\textrm{CB}}{l_\textrm{CA}}\right)
\,F_\textrm{max}
\quad\wedge\quad P = F_\textrm{BD} - F_\textrm{AE}
\end{equation*}

```python
FAEp = (lCB/lCA)*Fmax
print(f'FAE (FBDmax - plastifição) = {FAEp:6.2f} N')

Pp = Fmax - FAEp
print(f'P (plastificação da barra BD) = {Pp:6.2f} N')
```

Diagrama esquemático da história da deformada da estrutura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au10/Au10_P001_i5.png"
width="550">

Como a barra AE continua no domínio elástico, o alongamento axial sofrido pela barra é:

\begin{equation*}
\delta_A = \frac{F_\textrm{AE} l_\textrm{AE}}{AE}
\end{equation*}

```python
dAp = FAEp*lAE/AE
print(f'desl A (plastifição) = {dAp:6.3f} mm')
```

Da relação de compatibilidade de deslocamentos:

\begin{equation*}
\frac{\delta_\textrm{A}}{x}
= \frac{\delta_\textrm{C}}{l_\textrm{AC} - x}
\quad\Leftrightarrow\quad
x = \frac{\delta_\textrm{A}}{\delta_\textrm{A} + \delta_\textrm{C}}
l_\textrm{AC}
\qquad\wedge\qquad
\theta_f = \arctan\left(\frac{\delta_\textrm{A}}{x}\right)
\end{equation*}

```python
xp = dAp/(dAp + dC)*lCA
print(f'x (plastifição) = {xp:6.2f} mm')

rotFinal = np.rad2deg(np.arctan(dAp/xp))
print(f'rotação final = {rotFinal:6.3f} graus')
```

Em consequência da equação de compatibilidade de deslocamentos pode-se calcular o deslocamento efetivo axial da barra plastificada BD:

\begin{equation*}
\frac{\delta_\textrm{A}}{x}
= \frac{\delta_\textrm{B}}{l_\textrm{AB} - x}
\quad\Leftrightarrow\quad
\delta_\textrm{B} =
\frac{l_\textrm{AB} - x}{x} \delta_\textrm{A}
\end{equation*}

```python
dBtot = (lAB-xp)/xp*dAp
print(f'desl. B (plastifição) = {dBtot:6.3f} mm')
```

Incluindo também o alongamento da barra BD no momento da cedência:

\begin{equation*}
\delta_\textrm{B} = \frac{F_\textrm{BD} l_\textrm{BD} } {AE}
\quad\text{(no limite do domónio elástico)}
\end{equation*}

Obtêm-se a seguinte relação força-deslocamento para a barra BD:

```python
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 20})
# use LaTeX fonts in the plot
# plt.rc('text', usetex=True)
plt.rc('font', family='serif')

FBCmax = Fmax
print(f'Fmax barra BD = {FBCmax*1e-3:.2f} kN')

dBelas = Fmax*lBD/E/A
print(f'Desl. max B (Fmax) = {dBelas:.4f} mm')

x1, x2, y = dBelas, dBtot, FBCmax*1e-3
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot(1, 1, 1)
ax.plot([0,x1 ,2*x2], [0,y,y], color='tab:red', label='Barra BD')
ax.plot(x1,y,marker='o',color='r', markersize=10)
plt.text(x1*.71, y*1.05, f'({x1:.3f},{y})', fontsize=14)
ax.plot(x2,y,marker='o',color='r', markersize=10)
plt.text(x2*.81, y*.92, f'({x2:.3f},{y})', fontsize=14)
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0, fontsize=10)
plt.xlabel('Deslocamento, mm', fontsize=16)
plt.ylabel('Força, kN', fontsize=16)
ax.set_xlim(0,1.6*x2)
ax.set_ylim(0,y*1.2)
# Change the size of the axis ticks
ax.tick_params(axis='both', which='major', labelsize=14, length=6, width=1)
ax.tick_params(axis='both', which='minor', labelsize=10, length=4, width=0.5)

plt.grid()
plt.show()
```

De notar que se a força $P$ for removida a estrutura não recupera a sua configuração inicial pois a barra BD deformou plasticamente. A deformação plástica residual da barra BD é dada por:

\begin{equation*}
\delta_\textrm{B}^P = \delta_\textrm{B,tot} - \delta_\textrm{B,elas}
\end{equation*}

```python
dBresidual = dBtot - dBelas
print(f'deslo. B residual (plastifição:barra BD) = {dBresidual:6.3f} mm')
```

### c) Considere agora que essa força era retirada e calcule o deslocamento residual da extremidade C. <a id="C"></a> ([_enunciatum_](#PIV101))

Quando a força $P$ é removida a barra AE recupera completamente porque se deforma em regime elástico, contudo a barra BD tem uma deformação residual permanente. Esta deformação será responsável por uma configuração da estrutura diferente da inicial. Nomeadamente, para além do deslocamento residual no ponto B (pela deformação plástica da barra BD),  existirá um deslocamento no ponto C devido à rotação da barra rigida ABC em torno do ponto A. Neste cenário a equação de compatibilidade de deslocamentos pela geometria do problema é:

\begin{equation*}
\frac{\delta_\textrm{B}^P}{l_\textrm{AB}}
= \frac{\delta_\textrm{C}}{l_\textrm{AC}}
\quad\Leftrightarrow\quad
\delta_\textrm{C} = \left(\frac{l_\textrm{AC}}{l_\textrm{AB}}\right)
\delta_\textrm{B}^P
\qquad\wedge\qquad
\theta_r = \arctan\left(\frac{\delta_\textrm{C}}{l_\textrm{AC}}\right)
\end{equation*}

```python
dCresidual = (lCA/lAB)*dBresidual
print(f'desl. C residual (após remoção da força P) = {dCresidual:6.3f} mm')

rotResifual = np.rad2deg(np.arctan(dCresidual/lCA))
print(f'rotação residual = {rotResifual:6.3f} graus')
```
