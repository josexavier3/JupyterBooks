## Problema 6 <a name="PI16"></a>

Considere o seguinte tensor (kPa):

\begin{equation*}
\sigma_{ij} =
\begin{bmatrix}
x^2y & x(1-y^2) & 0 \\
x(1-y^2) & \displaystyle\frac{1}{3}(y^3-3y) & 0 \\
0 & 0 & 2z^3
\end{bmatrix}
\end{equation*}

Calcule as forças mássicas tal que as equações de equilíbrio sejam satisfeitas em todo o domínio.

## Resolução

A equação do movimento de um méio contínuo é descrita por (na ausência de forças de inércia: $ma_i$):

\begin{equation*}
\sigma_{ji,j} + \rho b_i = 0
\end{equation*}

Esta equação diferencial representa o equilíbrio de um ponto material
$\vec{x}(t)\in O(t)$. Esta pode ainda ser reescrita:

\begin{equation*}
\textrm{div} \tilde{\sigma}^T(\vec{x},t) + \vec{f}(\vec{x},t) =
\displaystyle\frac{\partial \sigma_{ji}}{\partial x_j} \vec{e}_j + f_j \vec{e}_j = \vec{0}
\end{equation*}

ou na forma matricial:

\begin{equation*}
\begin{bmatrix}
\displaystyle\frac{\partial \sigma_{xx}}{\partial x} +
\displaystyle\frac{\partial \sigma_{yx}}{\partial y} +
\displaystyle\frac{\partial \sigma_{zx}}{\partial z} \\
\displaystyle\frac{\partial \sigma_{xy}}{\partial x} +
\displaystyle\frac{\partial \sigma_{yy}}{\partial y} +
\displaystyle\frac{\partial \sigma_{zy}}{\partial z} \\
\displaystyle\frac{\partial \sigma_{xz}}{\partial x} +
\displaystyle\frac{\partial \sigma_{yz}}{\partial y} +
\displaystyle\frac{\partial \sigma_{zz}}{\partial z}
\end{bmatrix} +
\begin{bmatrix}
f_1 \\
f_2 \\
f_3 \\
\end{bmatrix}
= \begin{bmatrix}
0 \\
0 \\
0 \\
\end{bmatrix}
\end{equation*}

Esta equação é a representação matricial da 1ª lei de Cauchy do movimento dum meio contínuo, na base $\vec{e}_i$.


```python
%matplotlib inline
from sympy import symbols, diff
from sympy.solvers import solve

x, y, z, sxx, syy, szz, sxy, syx, syz, szy, sxz, szx, f1, f2, f3 = \
    symbols('x y z sxx syy szz sxy syx syz szy sxz szx f1 f2 f3')

def sxx(x,y): return x**2*y
def syy(x,y): return (y**3 -3*y)/3
def szz(x,y): return 2*z**3
def sxy(x,y): return x*(1-y**2)
def syx(x,y): return sxy(x,y)
# sigma_ij = sigma_ji - simetria do tensor das tensões
def syz(x,y): return 0
def szy(x,y): return syz(x,y)
def sxz(x,y): return 0
def szx(x,y): return sxz(x,y)

eq1 = diff(sxx(x,y), x) + diff(syx(x,y), y) + diff(szx(x,y), z) + f1
eq2 = diff(sxy(x,y), x) + diff(syy(x,y), y) + diff(szy(x,y), z) + f2
eq3 = diff(sxz(x,y), x) + diff(syz(x,y), y) + diff(szz(x,y), z) + f3

print('eq1 = (',diff(sxx(x,y), x),') + (',diff(syx(x,y), y),') + (',diff(szx(x,y), z),') + ',f1)
print('eq2 = (',diff(sxy(x,y), x),') + (',diff(syy(x,y), y),') + (',diff(szy(x,y), z),') + ',f2)
print('eq3 = (',diff(sxz(x,y), x),') + (',diff(syz(x,y), y),') + (',diff(szz(x,y), z),') + ',f3)

sols = solve((eq1, eq2, eq3),{f1, f2, f3}, dict=True)
print('solve: ', str(sols[0]))
F1 = sols[0][f1]
F2 = sols[0][f2]
F3 = sols[0][f3]

print('\nPelo que a força mássica bi (N//kg) vem : ')
print('f1 = rho b_x = ', F1)
print('f2 = rho b_y = ', F2)
print('f3 = rho b_z = ', F3)
```
