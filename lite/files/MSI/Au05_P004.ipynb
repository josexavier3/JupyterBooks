## Problema 4 <a name="PII54"></a>

O coeficiente de extensão volumétrica de um corpo é dado por $\Delta V/V$. Assumindo extensões
infinitesimais mostre que a extensão volumétrica é igual ao primeiro invariante das extensões.

### Nomenclatura

\begin{equation*}
\begin{aligned}
\vec{V} &\quad \textrm{volume inicial do corpo não deformado} \\
\vec{V}' &\quad \textrm{volume final do corpo deformado} \\
\Delta V &\quad \textrm{variação de volume entre a configuração inicial e atual} \\
\varepsilon_\mathrm{v} &\quad \textrm{coeficiente de extensão volumétrico ou dilatação} \\
\mathrm{d}\alpha &\quad \textrm{comprimento infinitesimal de um cubo elementar na direção $\alpha$} \\
&\quad \textrm{($\alpha = x,y,z$), na configuração de referência} \\
\mathrm{d}\alpha' &\quad \textrm{comprimento infinitesimal de um cubo elementar na direção $\alpha'$} \\
&\quad \textrm{($\alpha' = x',y',z'$), na configuração atual} \\
I_\mathrm{\varepsilon} &\quad \textrm{primeiro invariante do tensor das deformações)} \\
\end{aligned}
\end{equation*}

## Resolução

O coeficiente de extensão volumétrico ou dilatação é dado por:

\begin{equation*}
\varepsilon_\mathrm{v} = \frac{\Delta V}{V_0} = \frac{V'-V_0}{V_0}
\end{equation*}

em que $V'$ é o volume final do corpo deformado, $V_0$ é o volume inicial do corpo não deformado.

Um volume infinitesimal de um cubo de arestas $\mathrm{d}x$, $\mathrm{d}y$ e $\mathrm{d}z$ é determinado por:

\begin{equation*}
V_0 = \mathrm{d}x\mathrm{d}y\mathrm{d}z
\end{equation*}

A variação de comprimento das arestas do cubo é diretamente proporcinal à deformação linear nas direções dos eixos do sistema de coordenadas $\mathrm{d}i(1+\varepsilon_{ii}) ~(i=x,y,z)$,  devido à deformação do corpo vem:

\begin{equation*}
\begin{aligned}
V' &= \mathrm{d}x'\mathrm{d}y'\mathrm{d}z' = \mathrm{d}x(1+\varepsilon_{xx})\mathrm{d}y(1+\varepsilon_{yy})\mathrm{d}z(1+\varepsilon_{zz}) \\
&=(1+\varepsilon_{xx})(1+\varepsilon_{yy})(1+\varepsilon_{zz})\mathrm{d}x\mathrm{d}y\mathrm{d}z \\
&= (1+\varepsilon_{xx})(1+\varepsilon_{yy})(1+\varepsilon_{zz})V_0 \\
\end{aligned}
\end{equation*}


```python
import numpy as np
import sympy as sy
from sympy import diff, Derivative
from sympy.solvers import solve
from sympy import symbols

eX, eY, eZ = symbols('eX eY eZ')
sy.expand((1 + eX)*(1 + eY)*(1 + eZ))
```

\begin{equation*}
\begin{aligned}
V' &= (1+\varepsilon_{xx}+\varepsilon_{yy}+\varepsilon_{zz} + \varepsilon_{xx}\varepsilon_{yy} +
\varepsilon_{yy}\varepsilon_{zz} + \varepsilon_{xx}\varepsilon_{zz} + \varepsilon_{xx}\varepsilon_{yy}\varepsilon_{zz} )V_0 \\
\end{aligned}
\end{equation*}

Atendendo a que as extensões longitudinais também são infinitesimais, os produtos destas extensões são infinitesimais de ordem superior, pelo que podem ser desprezáveis. Então:

\begin{equation*}
\begin{aligned}
V' &= (1+\varepsilon_{xx}+\varepsilon_{yy}+\varepsilon_{zz} )V_0 \\
\end{aligned}
\end{equation*}

Desenvolvendo,

\begin{equation*}
\begin{aligned}
\frac{V'}{V} -1 &= \varepsilon_{xx}+\varepsilon_{yy}+\varepsilon_{zz} \\
\frac{V'-V}{V}&= \varepsilon_{xx}+\varepsilon_{yy}+\varepsilon_{zz}
\end{aligned}
\end{equation*}

logo,

\begin{equation*}
\begin{aligned}
\varepsilon_\mathrm{v} = I_\mathrm{\varepsilon}
\end{aligned}
\end{equation*}

em que $I_\mathrm{\varepsilon}$ é o primeiro invariante do tensor das deformações.
