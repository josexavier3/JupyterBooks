#### a) Determine o tensor das tensões no referencial $x_1x_2x_3$ em relação ao qual se obtiveram os quatro versores e os quatro vectores indicados.

([_enunciatum_](#PI41))

Uma mudança de base ou eixos de um sistema de coordenadas implica alterações nas componentes do tensor das tensões, expresso na matriz simétrica das tensões. Considere-se conhecida a matriz do tensor das tensões no sistema de coordenadas $S_0(0,\vec{\imath}_0,\vec{\jmath}_0,\vec{k}_0)$. Determine-se as suas componentes num novo referencial $S_1(0,\vec{\imath}_1,\vec{\jmath}_1,\vec{k}_1)$. Os cossenos diretores entre os dois sistemas de coordenadas é assim definido:

\begin{equation*}
\begin{array}{c|ccc}
& x_0 & y_0 & z_0 \\ \hline
x_1 & l_1  & m_1 & n_1 \\
y_1 & l_2  & m_2 & n_2 \\
z_1 & l_3  & m_3 & n_3 \\\hline
\end{array} \qquad \textrm{em que} \qquad
[Q] =
\begin{bmatrix}
l_1  & m_1 & n_1 \\
l_2  & m_2 & n_2 \\
l_3  & m_3 & n_3
\end{bmatrix}
\end{equation*}

em que a matriz $[Q] \equiv [Q]_{01}$ representa a matriz de transformação de $S_0$ em $S_1$.

Uma vez que a transformação de coordenadas é ortogonal, os cosenos directores estão relacionados pelas seguintes equações:

\begin{equation*}
\begin{aligned}
l_1^2+m_1^2 + n_1^2 &= 1 \quad & l_2^2+m_2^2 + n_2^2 &= 1 \quad & l_2^2+m_2^2 + n_3^2 &= 1 \\
l_1^2+l_2^2 + l_3^2 &= 1 \quad & m_1^2+m_2^2 + m_3^2 &= 1 \quad & n_1^2+n_2^2 + n_3^2 &= 1 \\
l_1l_2+m_1m_2+n_1n_2 &= 0 \quad & l_1l_3+m_1m_3+n_1n_3 &= 0 \quad & l_2l_3+m_2m_3+n_2n_3 &= 0 \\
l_1m_2+l_2m_2+l_3m_3 &= 0 \quad & l_1n_1+l_2n_2+l_3n_3 &= 0 \quad & m_1n_1+m_2n_2+m_3n_3 &= 0 \\
\end{aligned}
\end{equation*}

Pode concluir-se que: $\vec{n} \perp \vec{v} \perp \vec{m}$. As componentes destes versores no referencial $S(0,x,y,z)$ são:

\begin{equation*}
\begin{aligned}
\vec{v} &= (1) \vec{\imath} + (0) \vec{\jmath} + (0) \vec{k} \\
\vec{n} &= (0) \vec{\imath} + (0.866) \vec{\jmath} + (0.5) \vec{k} \\
\vec{m} &= (0) \vec{\imath} + (-0.5) \vec{\jmath} + (0.866) \vec{k}
\end{aligned} \quad \quad \quad
\begin{aligned}
\begin{bmatrix}
\vec{v} \\ \vec{n} \\ \vec{m}
\end{bmatrix} &=
\begin{bmatrix}
Q_{01}
\end{bmatrix}
\begin{bmatrix}
\vec{\imath} \\ \vec{\jmath} \\ \vec{k}
\end{bmatrix} \\
&=
\begin{bmatrix}
l_1  & m_1 & n_1 \\
l_2  & m_2 & n_2 \\
l_3  & m_3 & n_3
\end{bmatrix}
\begin{bmatrix}
\vec{\imath} \\ \vec{\jmath} \\ \vec{k}
\end{bmatrix} \\
&=
\begin{bmatrix}
1  & 0 & 0 \\
0  & 0.866 & 0.5 \\
0  & -0.5 & 0.866
\end{bmatrix}
\begin{bmatrix}
\vec{\imath} \\ \vec{\jmath} \\ \vec{k}
\end{bmatrix}
\end{aligned}
\end{equation*}

```python
import numpy as np
np.set_printoptions(precision=3, suppress=True)

def versor(v1,v2,v3): return np.transpose(np.array([v1,v2,v3]))

n = versor(0.0, 0.866, 0.500)
s = versor(0.0, 0.500, 0.866)
v = versor(1.0, 0.0, 0.0)
m = versor(0.0, -0.5, 0.866)

def angulosCosDir(vec,stg):
    l = np.arccos(vec[0])
    m = np.arccos(vec[1])
    n = np.arccos(vec[2])
    print('Versor : ',stg,' : ----------------------------------')
    print('l_',stg,' = ','{:.2f}'.format(np.rad2deg(l)),' Deg | m_',stg,' = ','{:.2f}'.format(np.rad2deg(m)),' Deg |  n_',stg,' = ','{:.2f}'.format(np.rad2deg(n)),' Deg')

angulosCosDir(n,'n')
angulosCosDir(s,'s')
angulosCosDir(v,'v')
angulosCosDir(m,'m')

nv = np.dot(n,v)
print('\nproduto escalar (n,v)',nv)
nm = np.dot(n,m)
print('produto escalar (n,m)',nm)
vm = np.dot(v,m)
print('produto escalar (v,m)',vm)

ns = np.dot(n,s)
print('\nproduto escalar (n,s)', '{:.3f}'.format(ns))
sm = np.dot(s,m)
print('produto escalar (s,m)', '{:.3f}'.format(sm))

print('\nversores v, n, e m formam um sistema de coordenadas ortonormado'
      ' e direto!')
def vetorTens(t1,t2,t3): return np.transpose(np.array([t1,t2,t3]))
```

As componentes da tensão $\vec{t}{(\vec{n})}$ no sistema de coordenadas $S_0(^0\underline{t})$ e no sistema de
coordenadas $S_1(^1\underline{t})$ estão relacionadas através das seguintes expressões:

\begin{equation*}
\underline{t}_0\,= \underline{\underline{Q}}_{10}\,\underline{t}_1
\quad\textrm{e}\quad
\underline{t}_1\,= \underline{\underline{Q}}_{01}\,\underline{t}_0
\end{equation*}

A matriz das tensões projetada em sistemas de coordenadas diferentes, $S_0$ e $S_1$ está relacionada pela seguinte lei
de transformação de um tensor de segunda ordem:

\begin{equation*}
\begin{aligned}
\underline{\underline{\sigma}}_1&=
\underline{\underline{Q}}_{01}\,\underline{\underline{\sigma}}_0~
\underline{\underline{Q}}_{01}^T
\end{aligned}
\end{equation*}

ou, pela transformação inversa,

\begin{equation*}
\begin{aligned}
\underline{\underline{\sigma}}_0&=
\underline{\underline{Q}}_{01}^T\,\underline{\underline{\sigma}}_1~
\underline{\underline{Q}}_{01} = \underline{\underline{Q}}_{10}\,\underline{\underline{\sigma}}_1~
\underline{\underline{Q}}_{10}^T
\end{aligned}
\end{equation*}

em que,

\begin{equation*}
\begin{aligned}
\underline{\underline{Q}}_{01}^T =
\underline{\underline{Q}}_{10}
\end{aligned}
\end{equation*}

```python
tv0 =  vetorTens(12.000, 0.000, 6.000)
print('\nvetor tensão (v) S1 = ',tv0)
tn0 =  vetorTens(3.000, 5.196, 2.000)
print('vetor tensão (n) S1 = ',tn0)
tm0 =  vetorTens(5.196, -3.000, 3.464)
print('vetor tensão (m) S1 = ',tm0)

def Q(a1,a2,a3): return np.block([[a1],[a2],[a3]])

Q01 = Q(v,n,m)
print('\nMatriz de transformação Q01 = \n',Q01)

Q10 = np.transpose(Q01)
print('\nMatriz de transformação Q10 = \n',Q10)

tv1 = np.dot(Q01,tv0)
print('\nVetor tensão (v) S1 = ',tv1)
tn1 = np.dot(Q01,tn0)
print('Vetor tensão (n) S1 = ',tn1)
tm1 = np.dot(Q01,tm0)
print('Vetor tensão (m) S1 = ',tm1)

def matrizTensao(vt1,vt2,vt3):
    return np.block([[vt1],[vt2],[vt3]])

tt1 = matrizTensao(tv1,tn1,tm1)
print('\nMatriz das tensões tt1 = \n',tt1)

tt0 = np.dot(Q10,np.dot(tt1,np.transpose(Q10)))
tt0 = np.around(tt0,decimals=6)
print('\nMatriz das tensões tt0 =  = Q10^T tt1 Q10^T = \n',tt0)

tt0_v2 = np.dot(np.transpose(Q01),np.dot(tt1,Q01))
print('\nMatriz das tensões tt0 = Q01^T tt1 Q01 = \n',tt0_v2)
```

#### b) Calcule os invariantes para esse tensor das tensões.<a name="B"></a>
([_enunciatum_](#PI41))

```python
def Invariantes(mt):
    inv1 = np.trace(mt)
    inv2 = 0.5*(np.trace(mt)**2 -np.trace(np.dot(mt,mt)))
    inv3 = np.linalg.det(tt0)
    return inv1, inv2, inv3
print('==================================================')
print('Pela matriz das tensões expressa no referencial S0:')
print('==================================================')
I1, I2, I3 = Invariantes(tt0)
print('1º Invariante, I1 (S0):', '{:.1f}'.format(I1))
print('2º Invariante, I2 (S0):', '{:.1f}'.format(I2))
print('3º Invariante, I3 (S0):', '{:.1f}'.format(I3))

print('==================================================')
print('Pela matriz das tensões expressa no referencial S1:')
print('==================================================')
I1_S1, I2_S1, I3_S1 = Invariantes(tt1)
print('1º Invariante, I1 (S1):', '{:.1f}'.format(I1_S1))
print('2º Invariante, I2 (S1):', '{:.1f}'.format(I2_S1))
print('3º Invariante, I3 (S1):', '{:.1f}'.format(I3_S1))

print('\n===================')
print('Tensões principais :')
print('===================')
valpropriosN = np.roots([1,-I1,I2,-I3])
print('Numpy : roots = ',valpropriosN)
valpropriosN = np.sort(valpropriosN)
sIII = valpropriosN[0]
sII = valpropriosN[1]
sI = valpropriosN[2]
print('\nTensão principal (valor próprio) sI ','{:.2f}'.format(sI))
print('Tensão principal (valor próprio) sII ','{:.2f}'.format(sII))
print('Tensão principal (valor próprio) sIII ','{:.2f}'.format(sIII))

print('\n================================')
print('Direcões principais (CALC 1 ) :')
print('================================')

import sympy as sy
from sympy.solvers import solve
from sympy import symbols

d1,d2,d3 = symbols('d1 d2 d3')

def eq1(d1,d2,d3,vp): return (tt0[0,0] - vp)*d1 + tt0[0,1]*d2 + tt0[0,2]*d3
def eq2(d1,d2,d3,vp): return tt0[1,0]*d1 + (tt0[1,1] - vp)*d2 + tt0[1,2]*d3
def eq3(d1,d2,d3,vp): return tt0[2,0]*d1 + tt0[2,1]*d2 + (tt0[2,2] - vp)*d3
def eq4(d1,d2,d3): return d1**2 + d2**2 + d3**2 - 1

def versor(e1,e2,e3): return np.array([e1,e2,e3])

# 1 valor proprio -------------------
print('\nVetor próprio associado a sI = ','{:.2f}'.format(sI),'.............................')
print('\n(i) system of equations:')
E1_sI = eq1(d1,d2,d3,sI); print(E1_sI,' = 0')
E2_sI = eq2(d1,d2,d3,sI); print(E2_sI,' = 0')
E3_sI = eq3(d1,d2,d3,sI); print(E3_sI,' = 0')
E4_sI = eq4(d1,d2,d3); print(E4_sI,' = 0')
VetProp1 = solve([E1_sI,E2_sI,E4_sI], [d1,d2,d3], dict=True)
l1 = float(VetProp1[1][d1])
m1 = float(VetProp1[1][d2])
n1 = float(VetProp1[1][d3])
versorI = versor(l1,m1,n1)
print('\n (iii) vec{nI} = ','{:.3f}'.format(l1),'e1 +','{:.3f}'.format(m1),'e2 +','{:.3f}'.format(n1),'e3')

# 2 valor proprio -------------------
print('\nVetor próprio associado a sII= ','{:.2f}'.format(sII),' .............................')
print('\n (i) system of equations:')
E1_sII = eq1(d1,d2,d3,sII); print(E1_sII,' = 0')
E2_sII = eq2(d1,d2,d3,sII); print(E2_sII,' = 0')
E3_sII = eq3(d1,d2,d3,sII); print(E3_sII,' = 0')
E4_sII = eq4(d1,d2,d3); print(E4_sII,' = 0')
VetProp2 = solve([E1_sII,E2_sII,E4_sII], [d1,d2,d3], dict=True)
l2 = float(VetProp2[1][d1])
m2 = float(VetProp2[1][d2])
n2 = float(VetProp2[1][d3])
versorII = versor(l2,m2,n2)
print('\n (iii) vec{nII} = (','{:.3f}'.format(l2),') e1 + (','{:.3f}'.format(m2),') e2 + (','{:.3f}'.format(n2),') e3')

# 3 valor proprio -------------------
print('\nVetor próprio associado a sIII= ','{:.2f}'.format(sIII),' .............................')
print('\n (i) system of equations:')
E1_sIII = eq1(d1,d2,d3,sIII); print(E1_sIII,' = 0')
E2_sIII = eq2(d1,d2,d3,sIII); print(E2_sIII,' = 0')
E3_sIII = eq3(d1,d2,d3,sIII); print(E3_sIII,' = 0')
E4_sIII = eq4(d1,d2,d3); print(E4_sIII,' = 0')
VetProp3 = solve([E1_sIII,E2_sIII,E4_sIII], [d1,d2,d3], dict=True)
l3 = float(VetProp3[0][d1])
m3 = float(VetProp3[0][d2])
n3 = float(VetProp3[0][d3])
versorIII = versor(l3,m3,n3)
print('\n (iii) vec{nIII} = (','{:.3f}'.format(l3),') e1 + (','{:.3f}'.format(m3),') e2 + (','{:.3f}'.format(n3),') e3')

print('\n Further verifications on the principal unit vectors: ')
print('dot(vp1,vp1) = ','{:.1f}'.format(np.dot(versorI,versorI)))
print('dot(vp2,vp2) = ', '{:.1f}'.format(np.dot(versorII,versorII)))
print('dot(vp3,vp3) = ', '{:.1f}'.format(np.dot(versorIII,versorIII)))
print('dot(vp1,vp2) = ', '{:.0f}'.format(np.abs(np.dot(versorI,versorII))))
print('dot(vp2,vp3) = ', '{:.0f}'.format(np.abs(np.dot(versorII,versorIII))))
print('dot(vp1,vp3) = ', '{:.0f}'.format(np.dot(versorI,versorIII)))

np.set_printoptions(precision=3)
print('\n Matriz de transformação de S0 para SI (ref. principal): ')

print('\n Àngulo entre eI e x: ')
print('theta_P = ','{:.3f}'.format(np.rad2deg(np.arccos(l1))), ' Deg')

mT = np.concatenate((versorI, versorII,versorIII))
mT = mT.reshape(3,3)
print('\n [T] = \n', mT)

print('\n[T][T]^T = I :\n',np.dot(np.transpose(mT),mT))

print('\n det(T) = ','{:.2f}'.format(np.linalg.det(mT)))

print('verificacao: ni.ni = 1 ')
eI2 = np.sqrt(np.dot(mT[0,:],mT[0,:]))
print('T[0,:] : eI2 = ','{:.2f}'.format(eI2))
eII2 = np.sqrt(np.dot(mT[1,:],mT[1,:]))
print('T[1,:] : eII2 = ','{:.2f}'.format(eII2))
eIII2 = np.sqrt(np.dot(mT[2,:],mT[2,:]))
print('T[2,:] : eII2 = ','{:.2f}'.format(eIII2))

print('\n================================')
print('Direcões principais (CALC 2 ) :')
print('================================')
def angP(s11,s22,s12): return np.arctan((2*s12)/(s11-s22))

sxx = tt0[0,0]
szz = tt0[2,2]
szy = tt0[0,2]

aP1 = angP(sxx,szz,szy)
aP2 = aP1 + np.pi

print('angulos das direções principais')
print('\nNo espaço das tensões:')
print('angPI  = ','{:.3f}'.format(np.rad2deg(aP1)),' deg | ','{:.3f}'.format(aP1),' radians')
print('angPII = ','{:.3f}'.format(np.rad2deg(aP2)),' deg | ','{:.3f}'.format(aP2),' radians')

aPI = aP1/2
aPII = aP2/2

print('\nNo espaço fisico:')
print('angPI  =  ','{:.3f}'.format(np.rad2deg(aPI)),' deg | ','{:.3f}'.format(aPI),' radians')
print('angPII = ','{:.3f}'.format(np.rad2deg(aPII)),' deg | ','{:.3f}'.format(aPII),' radians')

def Q(ang,rot):
    c = np.cos(ang)
    s = np.sin(ang)
    if rot == 1:
        Qij = np.array([[1,0,0],[0,c,s],[0,-s,c]])
    elif rot == 2:
        Qij = np.array([[c,0,s],[0,1,0],[-s,0,c]])
    else:
        Qij = np.array([[c,s,0],[-s,c,0],[0,0,1]])
    return Qij

MatrTrans = Q(aPI,2)
print('\nMatriz de transformação = \n',MatrTrans)

print('\nVetores principais:')
nI = MatrTrans[0,:]
print('eI = ', nI)
nII = MatrTrans[1,:]
print('eII = ', nII)
nIII = MatrTrans[2,:]
print('eIII = ', nIII)
```

#### c) Determine a tensão normal e tangencial no plano definido pelo versor $\vec{\lambda} = \begin{bmatrix}1/\sqrt{3} & 1/\sqrt{3} & 1/\sqrt{3}\end{bmatrix}$.<a name="C"></a> ([_enunciatum_](#PI41))

```python
verC = versor(1/np.sqrt(3),1/np.sqrt(3),1/np.sqrt(3))
print('Vetor = ',verC)
print('|| verC || = ',np.sqrt(np.dot(verC,verC)))

vetorTensaoC = np.dot(tt0,verC)
print('Vetor tensão = ',vetorTensaoC,' MPa')
sn = np.dot(vetorTensaoC,verC)
print('Tensão normal = ','{:.3f}'.format(sn),' MPa')
tau = np.sqrt(np.dot(vetorTensaoC,vetorTensaoC)-sn**2)
print('Tensão de corte  = ','{:.3f}'.format(tau),' MPa')
```

#### d) Verifique o resultado da alínea anterior utilizando o diagrama de Mohr.<a name="D"></a>

([_enunciatum_](#PI41))

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams, rc
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
# rc('text', usetex=True)

def mohr(S,limits):
    """Plot Mohr circle for a 2D tensor"""
    S11 = S[0][0]
    S12 = S[0][1]
    S22 = S[1][1]
    center = [(S11 + S22)/2.0, 0.0]
    radius = np.sqrt((S11 - S22)**2/4.0 + S12**2)
    Smin = center[0] - radius
    Smax = center[0] + radius

    print("  Tensão normal mínima: ", np.round(Smin,6))
    print("  Tensão normal máxima: ", np.round(Smax, 6))
    print("  Tensão normal média : ", np.round(center[0], 6))
    print("Tensão de corte mínima: ", np.round(-radius, 6))
    print("Tensão de corte máxima: ", np.round(radius, 6))

    print('Ponto A (','{:.1f}'.format(S11),',','{:.1f}'.format(S12),')')
    print('Ponto B (','{:.1f}'.format(S22),',','{:.1f}'.format(-S12),')')

    circ = plt.Circle((center[0],0), radius, facecolor='tab:olive', lw=3,
    edgecolor='tab:orange')
    desv = 0.1*radius
    plt.axis('image')
    ax = plt.gca()
    ax.add_artist(circ)
    ax.set_xlim(Smin - .3*radius, Smax + .3*radius)
    ax.set_ylim(-1.3*radius, 1.3*radius)
    plt.plot([S11, S22], [S12, -S12], 'ko')
    plt.plot([S11, S22], [S12, -S12], 'k')
    plt.plot(Smin, 0, 'bs')
    plt.plot(Smax, 0, 'bs')
    plt.plot(center[0], radius, 'rv')
    plt.plot(center[0], -radius, 'rv')
    plt.plot(center[0], center[1], 'o', mfc='w')
    plt.text(center[0]+desv, desv, 'C')
    plt.text(S11 + desv, S12 + desv , 'A')
    plt.text(S22 + desv, -S12 + desv, 'B')
    plt.text(Smin + .1*radius, -0.15*radius, r'$\sigma_\mathrm{II}$')
    plt.text(Smax - .2*radius, -0.15*radius, r'$\sigma_\mathrm{I}$')
    plt.text(center[0]-4*desv, 1.1*radius, r'$\tau_\mathrm{max}$')
    plt.text(center[0]-4*desv, -1.1*radius, r'$\tau_\mathrm{min}$')
    plt.xlabel(r"$\sigma$", size=18)
    plt.ylabel(r"$\tau$", size=18)
    plt.xlim(limits[0,0],limits[0,1])
    plt.ylim(limits[1,0],limits[1,1])
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    plt.show()

axislim = np.array([[0,16],[-10,10]])
print('\n================================')
print('PLANO x-z:')
print('================================')
sxx = tt0[0,0]
szz = tt0[2,2]
sxz = tt0[0,2]
S = np.array([[sxx, sxz],[sxz, szz]])
mohr(S, axislim)
```

```python
print('--------------------------')
print('Equações de transformação:')
print('--------------------------')

print('\nMatriz das tensões tt0 = \n',tt0)

def sxx1(sxx,syy,txy,t):
    return (sxx+syy)/2 + (sxx-syy)/2*np.cos(2*t) + txy*np.sin(2*t)

def syy1(sxx,syy,txy,t):
    return (sxx+syy)/2 - (sxx-syy)/2*np.cos(2*t) - txy*np.sin(2*t)

def txy1(sxx,syy,txy,t):
    return (syy-sxx)/2*np.sin(2*t) + txy*np.cos(2*t)

def angPrincipal(sxx,syy,txy):
    return np.arctan(2*txy/(sxx-syy))/2

def angS(s11,s22,s12): return -np.arctan((s11-s22)/(2*s12))

print('\nPLANO x-z:\n')
sxx = tt0[0,0] # neste caso: x = y; y = z
txy = tt0[0,2]
syy = tt0[2,2]

angt = np.arange(-45,145,1)
tsxx = sxx1(sxx,syy,txy,np.deg2rad(angt))
tsyy = syy1(sxx,syy,txy,np.deg2rad(angt))
ttxy = txy1(sxx,syy,txy,np.deg2rad(angt))

print('Tensões (normais) principais: -------------------------------------------')
aprinc = np.rad2deg(angPrincipal(sxx,syy,txy))
print('angPrincipal 1 (deg) =','{:.2f}'.format(aprinc))
tsxx1 = sxx1(sxx,syy,txy,np.deg2rad(aprinc))
tsyy1 = syy1(sxx,syy,txy,np.deg2rad(aprinc))
ttxy1 = txy1(sxx,syy,txy,np.deg2rad(aprinc))
print('tens xx plano principal [1] (MPa) =','{:.2f}'.format(tsxx1))
print('tens yy plano principal [1] (MPa) =','{:.2f}'.format(tsyy1))
print('tens xy plano principal [1] (MPa) =','{:.2f}'.format(ttxy1))

aprinc2 = aprinc + 90
print('\nangPrincipal 2 (deg) =','{:.2f}'.format(aprinc2))
tsxx2 = sxx1(sxx,syy,txy,np.deg2rad(aprinc2))
tsyy2 = syy1(sxx,syy,txy,np.deg2rad(aprinc2))
ttxy2 = txy1(sxx,syy,txy,np.deg2rad(aprinc2))
print('tens xx plano principal [2] (MPa) =','{:.2f}'.format(tsxx2))
print('tens yy plano principal [2] (MPa) =','{:.2f}'.format(tsyy2))
print('tens xy plano principal [2] (MPa) =','{:.2f}'.format(ttxy2))

print('\nTensões de corte maximas/mínima: -------------------------------------------')
aS1 = angS(sxx,syy,txy)/2 # ângulo no espaço físico, radianos
print('Ângulo tensão corte max (Deg) =','{:.2f}'.format(np.rad2deg(aS1)))
tcortexxMa = sxx1(sxx,syy,txy,aS1)
tcorteyyMa = syy1(sxx,syy,txy,aS1)
tcortexyMa = txy1(sxx,syy,txy,aS1)
print('tens xx - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcortexxMa))
print('tens yy - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcorteyyMa))
print('tens xy - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcortexyMa))

aS2 = aS1 + np.pi/2
print('Ângulo tensão corte min (Deg) =','{:.2f}'.format(np.rad2deg(aS2)))
tcortexxMb = sxx1(sxx,syy,txy,aS2)
tcorteyyMb = syy1(sxx,syy,txy,aS2)
tcortexyMb = txy1(sxx,syy,txy,aS2)
print('tens xx - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcortexxMb))
print('tens yy - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcorteyyMb))
print('tens xy - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcortexyMb))

# use LaTeX fonts in the plot
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(angt, tsxx, 'k--', label=r'$\sigma_{xx}$', linewidth=3)
ax.plot(angt, tsyy, 'k:', label=r'$\sigma_{yy}$', linewidth=3)
ax.plot(angt, ttxy, 'k', label=r'$\tau_{xy}$', linewidth=3)
# points: principal stresses
ax.plot(aprinc, tsxx1, 's'), ax.text(aprinc, tsxx1,r'$\sigma_{xx}$')
ax.plot(aprinc, tsyy1, 's'), ax.text(aprinc, tsyy1,r'$\sigma_{yy}$')
ax.plot(aprinc, ttxy1, 's'), ax.text(aprinc, ttxy1,r'$\tau_{xy}$')
ax.plot(aprinc2, tsxx2, 's'), ax.text(aprinc2, tsxx2,r'$\sigma_{xx}$')
ax.plot(aprinc2, tsyy2, 's'), ax.text(aprinc2, tsyy2,r'$\sigma_{yy}$')
ax.plot(aprinc2, ttxy2, 's'), ax.text(aprinc2, ttxy2,r'$\tau_{xy}$')
# points: critical shear stresses
ax.plot(np.rad2deg(aS1), tcortexxMa, 'v')
ax.text(np.rad2deg(aS1), tcortexxMa,r'$\varepsilon_{xx}$')
ax.plot(np.rad2deg(aS1), tcorteyyMa, 'v')
ax.text(np.rad2deg(aS1), tcorteyyMa,r'$\varepsilon_{yy}$')
ax.plot(np.rad2deg(aS1), tcortexyMa, 'v')
ax.text(np.rad2deg(aS1), tcortexyMa,r'$\gamma_{xy}$')
ax.plot(np.rad2deg(aS2), tcortexxMb, 'v')
ax.text(np.rad2deg(aS2), tcortexxMb,r'$\varepsilon_{xx}$')
ax.plot(np.rad2deg(aS2), tcorteyyMb, 'v')
ax.text(np.rad2deg(aS2), tcorteyyMb,r'$\varepsilon_{yy}$')
ax.plot(np.rad2deg(aS2), tcortexyMb, 'v')
ax.text(np.rad2deg(aS2), tcortexyMb,r'$\gamma_{xy}$')

ax.set(xlabel='Ângulo (deg)', ylabel='Transformação das tensões')
legend = ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., shadow=True, fontsize=18)
ax.axhline(y=0, color='k', linewidth=1)
ax.axvline(x=0, color='k', linewidth=1)
ax.grid()
plt.show()
```
