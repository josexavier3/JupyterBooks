## Problema 8 <a name="PIV98"></a>

A figura representa uma barra ACB colocada entre dois suportes rígidos A e B, sem quaisquer tensões à temperatura ambiente de 20 ºC.  A barra é construída em aço macio caracterizado por apresentar comportamento elastoplástico com $E = 200$ GPa e $\sigma_C$ = 250 MPa.  Considere que o coeficiente de dilatação do aço é $\alpha = 11,7 \times 10^{-6}$ /ºC e que a barra é submetida a um aumento de temperatura e depois arrefecida até regressar à temperatura ambiente. Sabendo que a tensão residual no troço CB da barra é de 60 MPa, determine qual a temperatura máxima atingida quando a barra foi aquecida.

width="700">


## Resolução

**Depois do arrefecimento**

\begin{equation*}
\sigma^R_\textrm{CB} = \frac{F^R}{A_\textrm{CB}}
\Leftrightarrow F^R = \sigma^R_\textrm{CB}A_\textrm{CB}
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# units: MPa, mm, N
E = 200.e9
sig_C = 250.e6
aplha_C = 11.7e-6
sig_R_CB = 60.e6
A_AC, A_CB = 320.e-6, 550.e-6
L_AC, L_CB = 0.250, 0.150
varDT = 20.

F_R = sig_R_CB*A_CB
print(f'F_R = {F_R:.1f} N')
```

\begin{equation*}
\delta^R = \frac{F^RL_\textrm{AC}}{A_\textrm{AC}E} + \frac{F^RL_\textrm{CB}}{A_\textrm{CB}E}
\end{equation*}

```python
delta_R = F_R*L_CB/A_CB/E + F_R*L_AC/A_AC/E
print(f'delta_R = {delta_R*1e3:.3f} mm')
```

Força residual instalada na barra depois do arrefecimento:

\begin{equation*}
F^R = F_\textrm{AC} = F_\textrm{CB} \quad \longleftarrow \quad \sum \vec{F} = \vec{0}
\end{equation*}

\begin{equation*}
F = F_\textrm{max} = \sigma_C A
\quad\wedge\quad
A_\textrm{AC} < A_\textrm{CB}
\end{equation*}


```python
Fmax = sig_C*A_AC
print(f'Fmax = {Fmax:.1f} N')
```

Resulta, portanto que o troço AC da barra plastifica:

\begin{equation*}
\delta^R = \delta^R_\textrm{AC}
\end{equation*}

**T_max (barra aquecida)**

\begin{equation*}
\delta_T = \delta_\textrm{AC}^e + \delta_\textrm{AC}^e + \delta^R
\end{equation*}

\begin{equation*}
\alpha\Delta T (L_\textrm{AC} + L_\textrm{CB}) = \delta_\textrm{AC}^e + \delta_\textrm{CB}^e + \delta^R_\textrm{AC}
\end{equation*}

```python
delta_AC_e = Fmax*L_AC/A_AC/E
print(f'delta_AC_e = {delta_AC_e*1e3:.3f} mm')
```

```python
delta_CB_e = Fmax*L_CB/A_CB/E
print(f'delta_CB_e = {delta_CB_e*1e3:.3f} mm')
```

```python
alpha_DT = aplha_C*(L_AC + L_CB)
DT_ = (delta_AC_e + delta_CB_e + delta_R)/alpha_DT
print(f'DT_ = {DT_:.1f} ºC')
```

```python
Tmax = 20. + DT_
print(f'Tmax = {Tmax:.1f} ºC')
```
