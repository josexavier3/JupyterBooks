## Problema 7 <a name="I1P7"></a>

Considere o seguinte estado de tensão (MPa):

\begin{equation*}
\underline{\underline{\sigma}}^\top =
\begin{bmatrix}
9 & 12 & 0 \\
12 & -9 & 0 \\
0 & 0 & 5
\end{bmatrix}
\end{equation*}

Calcule as componentes normal, $\sigma_n$, e tangencial, $\tau$, do vetor de tensão, $\vec{t}{\,(p,\vec{n})}$, no plano (faceta) normal ao versor $\vec{n} = 1/5(4\,\vec\imath + 3 \,\vec\jmath)$.

## Resolução

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def versor(n1,n2,n3):
    return np.array([n1, n2, n3])

def tensortensoes(s11, s22, s33, s23, s13, s12):
    return np.array([[s11,s12,s13], [s12,s22,s23], [s13,s23,s33]])

facetaPn = versor(4/5,3/5,0)
print('vec n = ', facetaPn)

# notacao de Voigt: S11, S22, S33, S23, S13, S12
ttens = tensortensoes(9,-9,5,0,0,12)
print('tensor sigma = ', ttens,'MPa')

tPn = np.dot(facetaPn,ttens)
print('vetor tensao = ', tPn, 'MPa')

sigman = np.dot(tPn,facetaPn)
print('tensao normal = ', '{:.3f}'.format(sigman),'MPa')

tau = np.sqrt(np.dot(tPn,tPn) - sigman**2)
print(f'tensao tangencial = {tau:.3f} MPa')
```
