#### a) Escreva o tensor das tensões no ponto $P$ para o referencial indicado.<a name="A"></a> ([_enunciatum_](#PV111))

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au11/Au11_P001_i2.jpg"
width="600" height="210">

- Tensão normal

\begin{equation*}
\sigma_{xx} = \frac{N}{A}
\end{equation*}

- Tensão de corte

\begin{equation*}
\tau_{xy} = \frac{T}{J/\rho}
\end{equation*}

em que, $T$ é o momento torsor aplicado (N.m), $\rho$ é a coordenada radial da posição do veio em relação ao seu centro (m), $J$ é o momento polar de inércia ou segundo momento polar de área de secção tranversal (m $^4$) - descrever a resistência de um objeto maciço ao movimento rotacional.

O momento torsor $T$ aplicado ao cilíndro produz tensões de corte $\tau$ em faces perpendiculares ao eixo do veio. Assume-se que para um veio circular submetido a um momento de torção, cada secção transversal permanece plana e não distorcida. A tensão de corte $\tau_{xy}$ varia linearmente com a coordenada $\rho$, medida em relação ao eixo do cilindro.

- Matrix do tensor das tensões no referencial $xyz$:

\begin{equation*}
\sigma_{ij} =
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} & \tau_{xz} \\
\tau_{xy} & \sigma_{yy} & \tau_{yz} \\
\tau_{xz} & \tau_{yz} & \sigma_{zz} \\
\end{bmatrix} =
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} & 0 \\
\tau_{xy} & 0 & 0 \\
0 & 0 & 0 \\
\end{bmatrix}~\quad\text{(MPa)}
\end{equation*}

Considerando a geometria do cilindro:

- Área de secção transversal e momento de inércia polar de área

\begin{equation*}
A^\bigcirc = \frac{\pi d^2}{4}
~~\text{(mm$^2$)}~;\quad
J^\bigcirc = \frac{\pi r^4}{2}
~~\text{(mm$^4$)}
\end{equation*}

em que,

\begin{equation*}
J^\bigcirc = \int_A\rho² \mathrm{d}A
\quad\wedge\quad\mathrm{d}A = 2\pi\rho\mathrm{d}\rho
\quad\therefore\quad
J^\bigcirc = 2\pi \int_0^r \rho³ \mathrm{d}\rho  = \frac{\pi r^4}{2}
\end{equation*}


descrever a resistência de um objeto maciço ao movimento rotacional, não sua resistência à deformação torcional

pelo que o tensor das tensões no ponto $P$ se pode escrever por:

\begin{equation*}
\sigma_{ij}(P) =
\begin{bmatrix}
4N/ \pi d^2 & 32T\rho/\pi d^4 & 0 \\
2T\rho/\pi r^4 & 0 & 0 \\
0 & 0 & 0 \\
\end{bmatrix}~\quad\text{(MPa)}
\end{equation*}

```python
import numpy as np
from typing import NamedTuple

# Dados
# unidades: m, Pa, N
class dados(NamedTuple):
    N : float # kN
    T : float # kN.m
    dia : float  # m
    rho : float 
    
d = dados(80, 15, 0.200, 0.080)

sxx = 4*d.N/np.pi/d.dia**2
print(f'Sxx = {sxx*1e-3:.3f} (MPa)')

sxy = 32*d.T*d.rho/np.pi/d.dia**4
print(f'Sxy = {sxy*1e-3:.3f} (MPa)')
```

#### b) Calcule as tensões principais e as direcções principais do estado de tensão nesse ponto.<a name="B"></a>


([enunciatum](#PV111))

- Método do círculo de Mohr

Centro da circunferência:

\begin{equation*}
\overline{\sigma} = \frac{\sigma_{xx}+\sigma_{yy}}{2}
= \frac{2N}{\pi d^2}
\end{equation*}

Raio da circunferência:

\begin{equation*}
R = \left[\left(\frac{2N}{\pi d^2}\right)^2 +
\left(\frac{32T\rho}{\pi d^4}\right)^2\right]^{1/2}
\end{equation*}

```python
print('Método do círculo de Mohr:-----------')
centro = sxx/2
print(f'centro = {centro*1e-3:.3f} (MPa)')

raio = (sxx**2/4 + sxy**2)**(1/2)
print(f'  raio = {raio*1e-3:.3f} (MPa)')

sI,sII, sIII = centro + raio, 0, centro - raio
print('\n Tensões principais:-----------')
print(f'  sI = {sI*1e-3:.3f} (MPa)')
print(f' sII = {sII*1e-3:.3f} (MPa)')
print(f'sIII = {sIII*1e-3:.3f} (MPa)')
```

As direções principais podem ser completamente obtidas neste caso pela definição do ângulo $\theta$, assim determinado:

\begin{equation*}
\tan2\theta_P = \frac{\tau_{xy}}{(\sigma_{xx} - \sigma_{yy})/2}
= \frac{2\tau_{xy}}{\sigma_{xx}} = 16 \frac{T\rho}{Nd²}
\end{equation*}

```python
twotheta = np.arctan(2*sxy/sxx)
twotheta2 = np.arctan(16*d.T*d.rho/d.N/d.dia**2)
print(f'theta (Mohr) = {twotheta:.3f} rad | {np.rad2deg(twotheta):.2f} º')
print(f'theta (físico) = {twotheta/2:.3f} rad | {np.rad2deg(twotheta/2):.2f} º')

def versor(n1,n2,n3): return np.array([n1,n2,n3])

aPI = twotheta/2
print('\nVetors principais:')
nI = versor(np.cos(aPI),np.cos(np.pi/2-aPI),round(np.cos(np.pi/2)))
print(nI)
nII = versor(round(np.cos(np.pi/2)),round(np.cos(np.pi/2)),np.cos(0))
print(nII)
nIII = versor(np.cos(np.pi/2+aPI),np.cos(aPI),round(np.cos(np.pi/2)))
print(nIII)
```

```python
def tensortens(sxx,sxy,sxz,syy,syz,szz):
    return np.array([[sxx,sxy,sxz],[sxy,syy,syz],[sxz,syz,szz]])

tt0 = tensortens(sxx*1e-3,sxy*1e-3,0,0,0,0) # unidades; kPa
print(f'tt0 = \n {tt0} (MPa)')
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams, rc
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 18
rc('text', usetex=False)
np.set_printoptions(precision=3)

print('--------------------------')
print('Circulo de Mohr')
print('--------------------------')

def mohr(S,limits):
    """Plot Mohr circle for a 2D tensor"""
    S11, S12, S22 = S[0][0], S[0][1], S[1][1]
    center = [(S11 + S22)/2.0, 0.0]
    radius = np.sqrt((S11 - S22)**2/4.0 + S12**2)
    Smin = center[0] - radius
    Smax = center[0] + radius

    print("Convenção- tensão de corte na faceta A com sentido anti-horário : - tau_{xy}")
    print(f'Ponto A ( {S11:.1f}, {-S12:.1f} )')
    print(f'Ponto B ( {S22:.1f}, {S12:.1f} )')

    print("  Tensão normal mínima: ", np.round(Smin,6))
    print("  Tensão normal máxima: ", np.round(Smax, 6))
    print("  Tensão normal média : ", np.round(center[0], 6))
    print("Tensão de corte mínima: ", np.round(-radius, 6))
    print("Tensão de corte máxima: ", np.round(radius, 6))

    circ = plt.Circle((center[0],0), radius, facecolor='#cce885', lw=3,
    edgecolor='#5c8037')
    desv = 0.1*radius
    plt.axis('image')
    ax = plt.gca()
    ax.add_artist(circ)
    ax.set_xlim(Smin - .3*radius, Smax + .3*radius)
    ax.set_ylim(-1.3*radius, 1.3*radius)
    plt.style.use('fast')
    plt.plot([S11, S22], [-S12, S12], 'ko')
    plt.plot([S11, S22], [-S12, S12], 'k')
    plt.plot(Smin, 0, 'bs')
    plt.plot(Smax, 0, 'bs')
    plt.plot(center[0], -radius, 'rv')
    plt.plot(center[0], radius, 'rv')
    plt.plot(center[0], center[1], 'o', mfc='w')
    plt.text(center[0]+desv, desv, 'C')
    plt.text(S11 + desv, -S12 , 'A')
    plt.text(S22 + desv, S12 , 'B')
    plt.text(Smin - 0.25*radius, -0.15*radius, r'$\sigma_\mathrm{min}$')
    plt.text(Smax + desv, -0.15*radius, r'$\sigma_\mathrm{max}$')
    plt.text(center[0]-4*desv, 1.1*radius, r'$\tau_\mathrm{max}$')
    plt.text(center[0]-4*desv, -1.1*radius, r'$\tau_\mathrm{min}$')
    plt.xlabel(r"$\sigma$", size=18)
    plt.ylabel(r"$\tau$", size=18)
    plt.xlim(limits[0,0],limits[0,1])
    plt.ylim(limits[1,0],limits[1,1])
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    plt.show()

S = tt0

axislim = np.array([[-10,15],[-10,10]])
print('\nPLANO x-z:\n')
S11,S12, S22 = S[0,0], S[0,1], S[1,1]
mohr(np.array([[S11, S12],[S12, S22]]),axislim)
```

```python
def circleradius(S):
    S11, S12, S22 = S[0,0], S[0,1], S[1,1]
    center = [(S11 + S22)/2.0, 0.0]
    radius = np.sqrt((S11 - S22)**2/4.0 + S12**2)
    return center, radius

limits = np.array([[-7.5,10.5],[-10,10]])

sigI, sigII, sigIII = sI*1e-3, sII*1e-3, sIII*1e-3

s11, s22, s12 = sigI, sigIII, 0
S = np.array([[s11, s12],[s12, s22]])
center1, radius1  = circleradius(S)

s11, s22, s12 = sigI, sigII, 0
S = np.array([[s11, s12],[s12, s22]])
center2, radius2  = circleradius(S)

s11, s22, s12 =  sigII, sigIII, 0
S = np.array([[s11, s12],[s12, s22]])
center3, radius3  = circleradius(S)

circ1 = plt.Circle((center1[0],0), radius1, facecolor='b')
circ2 = plt.Circle((center2[0],0), radius2, facecolor='r')
circ3 = plt.Circle((center3[0],0), radius3, facecolor='g')

plt.axis('image'); ax = plt.gca()

plt.gcf().gca().add_artist(circ1)
plt.gcf().gca().add_artist(circ2)
plt.gcf().gca().add_artist(circ3)

plt.xlabel(r"$\sigma$", size=18)
plt.ylabel(r"$\tau$", size=18)

plt.text(1.*sigI, 0, r'$\sigma_\mathrm{I}$')
plt.text(1.*sigII, 0, r'$\sigma_\mathrm{II}$')
plt.text(1.*sigIII, 0, r'$\sigma_\mathrm{III}$')

plt.xlim(limits[0,0],limits[0,1])
plt.ylim(limits[1,0],limits[1,1])

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

#### d) Determine analiticamente as tensões normal e tangencial no plano que passa em $P$ e cujo versor faz ângulos iguais com os três eixos do referencial..<a name="D"></a>

([enunciatum](#PV111))

O vetor que perfaz ângulos iguais com os três eixos do referencial é um vetor com iguais cosenos diretores: $\vec{v} = (1,1,1)$. O versor que define a direção desse vetor é unitário, i.e.: $||\vec{v}|| = n_1^2 + n_2^2 + n_3^2  = 1 \Leftrightarrow n = 1/\sqrt{3}$. Finalmente: $\vec{n} = \vec{v}/||\vec{v}||$.

A componente do vetor das tensões, $\vec{t}^{(\hat{n})}\equiv \vec{\sigma}^{(\vec{n})}$. segundo a normal à faceta designa-se por tensão normal, $\sigma_n$, e pode ser calculada pelo produto interno de $\vec{t}^{(\vec{n})}$ por $\vec{n}$:

\begin{equation*}
\sigma_n(p,\vec{n}) = \vec{t}^{(\vec{n})} \cdot \vec{n} = t_i^{(\vec{n})} \cdot n_i
\quad \text{;} \quad
\vec{\sigma}_n(p,\vec{n}) = (\vec{t}^{(\vec{n})} \cdot \vec{n})\vec{n}
\end{equation*}

A tensão normal diz-se de tração se for positiva e diz-se de compressão no caso negativo.

A componente de $\vec{t}^{(\vec{n})}$ no plano da faceta, designa-se por tensão tangencial. O cálculo da tensão tangencial pode ser obtido atendendo ao teorema de Pitágoras:

\begin{equation*}
||t_i^{(\vec{n})}||^2 = \sigma_n^2 + \tau^2
\end{equation*}

pelo que, o módulo da tensão tangencial vem,

\begin{equation*}
\tau(p,\vec{n}) = ||\vec{t}^{(\vec{n})} - \vec{\sigma}_n|| =
\sqrt{\vec{t}^{(\vec{n})}\cdot\vec{t}^{(\vec{n})} - \sigma_n^2} \quad \text{;} \quad
\vec{\tau}(p,\vec{n}) \equiv \tau\vec{s} = \vec{t} - (\vec{t}^{(\vec{n})} \cdot \vec{n})\vec{n}
\end{equation*}

```python
n = versor(1/np.sqrt(3),1/np.sqrt(3),1/np.sqrt(3))
print('normal à faceta = ', n)

print(f'ângulo n.i  = {np.rad2deg(np.arccos(1/np.sqrt(3))):.2f} º')

tPn = np.dot(tt0,n)
print(f'vetor tensao (MPa) = {tPn}')

sigman = np.dot(tPn,n)
print(f'tensao normal (MPa) = {sigman:.3f}')

tau = np.sqrt(np.dot(tPn,tPn) - sigman**2)
print(f'tensao tangencial (MPa) = {tau:.3f}')
```

```python
n = versor(1/np.sqrt(3),1/np.sqrt(3),1/np.sqrt(3))
print('normal à faceta = ', n)

print(f'ângulo n.i  = {np.rad2deg(np.arccos(1/np.sqrt(3))):.2f} º')

tPn = np.dot(tt0,n)
print(f'vetor tensao (MPa) = {tPn}')

sigman = np.dot(tPn,n)
print(f'tensao normal (MPa) = {sigman:.3f}')

tau = np.sqrt(np.dot(tPn,tPn) - sigman**2)
print(f'tensao tangencial (MPa) = {tau:.3f}')
```
