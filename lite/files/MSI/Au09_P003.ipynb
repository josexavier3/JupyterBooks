## Problema 3 <a name="PIV93"></a>

Um tubo de alumínio está colado a um núcleo de latão e o conjunto está livre de tensões a 15ºC. Considerando apenas tensões axiais calcule as tensões geradas no alumínio e no latão quando a temperatura atinge 100ºC. Os coeficientes de expansão do latão e do alumínio são respectivamente: $\alpha_{L}= 19\times 10^{-6}$/ºC e $\alpha_{AL}=23\times 10^{-6}$ /ºC. $E_\mathrm{alum}$= 70 GPa e $E_\mathrm{latão}$= 105 GPa. Diâmetro exterior do conjunto: 50 mm. Diâmetro do núcleo de latão: 25 mm.


O tudo de latão (interior) e o tubo de alumínio (exterior), solidários entre si, irão sobrer uma ditalação térmica conjunta, pela variação (aumento) de temperatura, passando a ocupar uma nova posição (configuração final). A variação de comprimento do conjunto pode ser analisado pela sobreposição de efeitos isolados. O alúminio irá sofrer um alongamento de origem térmic $\delta_{AL} = \alpha_{AL} \Delta TL$ enquando que o Latão irá sobrer uma variação térmica: $\delta_{L} = \alpha_{L} \Delta TL$. Uma vez que $\alpha_{AL} > \alpha_{L}$ (com $\Delta T$ e $L$ igual), o alumínio isoladamente poderia alongar-se mais do que o alongamento observado no latão. Contudo, porque existe compatibilidade de deslocamentos, existirá uma força exercída em ambos que alune essa diferença de alongamentos para a configuração final.

##### Equação de compatibilidade

\begin{equation*}
\begin{aligned}
\delta_{AL}^T - \delta_{AL}
= \delta_{L}^T + \delta_{L}
&~\Leftrightarrow~
\alpha_{AL} \Delta T L - \frac{FL}{A_{AL} E_{AL}}
= \alpha_{L} \Delta T L + \frac{FL}{A_{L} E_{L}} \\
&~\Leftrightarrow~
F= \frac{\Delta T  (\alpha_{AL}- \alpha_{L})}
{\left(\displaystyle\frac{1}{A_{AL}E_{AL}} + \displaystyle\frac{1}{A_{L}E_{L}} \right)}
\end{aligned}
\end{equation*}

```python
from numpy import pi

# unidades: mm, N, MPa
Eal, EL = 70.e3, 105.e3

de, di = 50., 25.

alphaal, alphaL = 23.e-6, 19.e-6

Ti, Tf = 15.,  100.
DT = Tf - Ti

Aal = pi*(de**2-di**2)/4
print(f'Aal = {Aal:.3f} mm2')

AL = pi*di**2/4
print(f'AL = {AL:.3f} mm2')

num = 1/Aal/Eal + 1/AL/EL
F = DT*(alphaal - alphaL)/num
print(f'F = {F:.3f} N')

Sigal = F/Aal
print(f'Sigal = {Sigal:.2f} MPa')

SigL = F/AL
print(f'SigL = {SigL:.2f} MPa')
```
