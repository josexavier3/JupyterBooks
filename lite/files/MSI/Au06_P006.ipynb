## Problema 6 <a name="PIII66"></a>

Um bloco de aço, de dimensões 80 $\times$ 60 $\times$ 40 mm $^3$,  está sujeito a uma pressão uniforme em todas as faces. Assuma $E = 200$ GPa e $\nu =$ 0.29. Sabendo que o lado maior teve uma variação
de comprimento de 24 microns calcule:

[a)](#A) A variação de comprimento das outras arestas.

[b)](#B) A pressão aplicada às faces do bloco.



## Resolução

#### a) A variação de comprimento das outras arestas.<a name="A"></a>

([enunciatum](#PIII66))

##### Da lei de Hooke para um material isotrópico.

Para um material isotrópico a lei de Hooke tridimensional vem expressa por:

\begin{equation*}
\varepsilon_{ij} = \frac{1}{E}[(1+\nu)\sigma_{ij} - \nu\delta_{ij}\sigma_{kk}]
\end{equation*}

Em forma matricial a lei constitutiva linear elástica expressa em termos de módulo de Young e coeficiente de Poisson
vem:

\begin{equation*}
\begin{aligned}
\varepsilon_{11} &= \displaystyle\frac{1}{E}[\sigma_{11} - \nu(\sigma_{22}+\sigma_{33})] \\
\varepsilon_{22} &= \displaystyle\frac{1}{E}[\sigma_{22} - \nu(\sigma_{11}+\sigma_{33})] \\
\varepsilon_{33} &= \displaystyle\frac{1}{E}[\sigma_{33} - \nu(\sigma_{11}+\sigma_{22})] \\
2\varepsilon_{23} &= \gamma_{23} = \displaystyle\frac{\tau_{23}}{G}\\
2\varepsilon_{13} &= \gamma_{13} = \displaystyle\frac{\tau_{13}}{G}\\
2\varepsilon_{12} &= \gamma_{12} = \displaystyle\frac{\tau_{12}}{G}\\
\end{aligned}
\end{equation*}

com,

\begin{equation*}
G = \frac{E}{2(1+\nu)}
\end{equation*}

Na forma matricial têm-se:

\begin{equation*}
\begin{bmatrix}
\varepsilon_{11} \\ \varepsilon_{22} \\ \varepsilon_{33} \\ 2\varepsilon_{23} \\ 2\varepsilon_{13} \\ 2\varepsilon_{12}
\end{bmatrix} = \frac{1}{E}
\begin{bmatrix}
1 & -\nu & -\nu & 0 & 0 & 0 \\
-\nu & 1 & -\nu & 0 & 0 & 0 \\
-\nu & -\nu & 1 & 0 & 0 & 0 \\
0 & 0 & 0 & 2(1+\nu) & 0 & 0 \\
0 & 0 & 0 & 0 & 2(1+\nu) & 0 \\
0 & 0 & 0 & 0 & 0 & 2(1+\nu) \\
\end{bmatrix}
\begin{bmatrix}
\sigma_{11} \\ \sigma_{22} \\ \sigma_{33} \\ \tau_{23} \\  \tau_{13} \\ \tau_{12}
\end{bmatrix}
\end{equation*}

A segunda lei da termodinâmica requer que as constantes elásticas verifiquem
os seguintes inequações:

\begin{equation*}
\begin{aligned}
E,~G > 0 \\
-1 < \nu < \frac{1}{2}
\end{aligned}
\end{equation*}

Para o problema proposto, o estado de tensão é dado por:

\begin{equation*}
\sigma_{ij} \equiv [\sigma]^\top =
\begin{bmatrix}
\sigma_{xx} & 0 & 0 \\
0 & \sigma_{yy} & 0 \\
0 & 0 & \sigma_{zz} \\
\end{bmatrix} =
\begin{bmatrix}
P & 0 & 0 \\
0 & P & 0 \\
0 & 0 & P \\
\end{bmatrix} =
P~\delta_{ij}~~~(i,j = 1,2,3) \quad\textrm{(MPa)}
\end{equation*}

As correspondentes componentes do tensor das deformações vêm:

\begin{equation*}
\begin{aligned}
\varepsilon_{xx} & = \varepsilon_{yy} = \varepsilon_{zz} =
\frac{(1-2\nu)}{E}P \\
\gamma_{yz} & = \gamma_{xz} = \gamma_{xy} = 0\\
\end{aligned}
\end{equation*}

- A variação de comprimento pode ser obtida por:

Por definição de extensão linear:

\begin{equation*}
\varepsilon_{kk} = \frac{\Delta L_k}{L_k}
\end{equation*}

A resolução do problema vem:

\begin{equation*}
\varepsilon_{xx} = \frac{\Delta L_x}{L_x} = \frac{(1-2\nu)}{E}P
\quad\Leftrightarrow\quad
P = \frac{E}{(1-2\nu)} \frac{\Delta L_x}{L_x}
\end{equation*}

```python
E = 200*1e3 # unit: MPa
v = 0.29
G = E/2/(1+v)
print('Propriedades do material: ---------------')
print('     E (MPa) = ',E)
print('           v = ',v)
print(f'     G (MPa) = {G:.2f}')

Lx, Ly, Lz = 80, 60, 40 # unit: mm
DLx = 24*1e-3 # unit: mm
exx = DLx/Lx
print('\n     exx = DLx\\Lx = ','{:.3e}'.format(exx))
P = E/(1-2*v)*exx
print('                P = ','{:.3f}'.format(P))
eyy = P*(1-2*v)/E
print('eyy = P*(1-2*v)\\E = ','{:.3e}'.format(eyy))
ezz = P*(1-2*v)/E
print('ezz = P*(1-2*v)\\E = ','{:.3e}'.format(ezz))
print('\n         DLx = ','{:.3f}'.format(DLx),' mm | ', '{:.1f}'.format(DLx*1e3),' micro')
DLy = eyy*Ly
print('DLy = eyy*Ly = ','{:.3f}'.format(DLy),' mm | ', '{:.1f}'.format(DLy*1e3),' micro')
DLz = ezz*Lz
print('DLz = ezz*Lz = ','{:.3f}'.format(DLz),' mm | ', '{:.1f}'.format(DLz*1e3),' micro')
```

#### b) A pressão aplicada às faces do bloco.<a name="B"></a>

([enunciatum](#PIII66))

\begin{equation*}
P = \frac{E}{(1-2\nu)} \frac{\Delta L_x}{L_x}
\end{equation*}

```python
P2 = E/(1-2*v)*(DLx/Lx)
print(f'P = {P2:.3f} MPa')
```
