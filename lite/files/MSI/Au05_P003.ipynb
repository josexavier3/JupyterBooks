## Problema 3 <a name="PII53"></a>

Um estado de deformação pode ser obtido a partir do seguinte campo de deslocamentos:

\begin{equation*}
	\begin{aligned}
	u &= Ax^2 + Bxy + C  \\
	v &= By^2 + Cx + Mz \\
	w &= Mz^2
	\end{aligned}
\end{equation*}

Investigue se são respeitadas as condições de compatibilidade das deformações.


## Resolução

### Equações de compatibilidade de Saint-Venant ou Equações de continuidade

O movimento de um ponto material de um objeto pode ser representado por um vetor deslocamento, $\underline{u}$. Este vetor pode ser expresso em função de três componentes cartesianos em relação a um dado sistema de coordenadas $S$. Além disso é possível espeficar um tensor de deformações no ponto em análise. A relação deslocamento-deformação de Cauchy exprime as seguintes componentes do tensor das deformações infinitésimais:

\begin{equation*}
\varepsilon_{ij} = \frac{1}{2}\left( \frac{\partial u_i}{\partial x_j}
+ \frac{\partial u_j}{\partial x_i} \right)
\end{equation*}

Explicitamente:

\begin{equation*}
\begin{aligned}
\varepsilon_{xx} &= \frac{\partial u}{\partial x} \\
\varepsilon_{yy} &= \frac{\partial v}{\partial y} \\
\varepsilon_{zz} &= \frac{\partial w}{\partial z} \\
\varepsilon_{xy} &= \frac{1}{2}\left( \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} \right) =  \frac{1}{2} \gamma_{xy} \\
\varepsilon_{xz} &= \frac{1}{2}\left( \frac{\partial u}{\partial z} + \frac{\partial w}{\partial x} \right) =  \frac{1}{2} \gamma_{xz}\\
\varepsilon_{yz} &= \frac{1}{2}\left( \frac{\partial v}{\partial z} + \frac{\partial w}{\partial y} \right) =  \frac{1}{2} \gamma_{yz}\\
\end{aligned}
\end{equation*}

A determinação das componentes da matriz do tensor das deformações a partir do campo de deslocamentos num ponto é direta
pois envolve apenas primeiras derivadas parciais.

```python
import numpy as np
import sympy as sy
from sympy import diff, Derivative
from sympy.solvers import solve
from sympy import symbols

u, v, w, x, y, z, A, B, C, M = symbols('u v w x y z A B C M')

def u(x,y,A,B,C): return A*x**2 + B*x*y + C
def v(x,y,z,B,C,M): return B*y**2 + C*x + M*z
def w(z,M): return M*z**2

u1 = u(x,y,A,B,C); print('desloc. em x : u  = ',u1)
v1 = v(x,y,z,B,C,M); print('desloc. em y : v = ',v1)
w1 = w(z,M); print('desloc. em z : w = ',w1)

# componentes cartesianas da matriz das deformações
exx = diff(u1, x)
print('\nexx = du\\dx :',exx)
eyy = diff(v1, y)
print('eyy = dv\\dy :',eyy)
ezz = diff(w1, z)
print('ezz = dw\\dz :',ezz)
exy  = 0.5*(diff(u1, y) + diff(v1, x) )
print('exy = 1\\2 gxy = 1\\2(du\\dy+dv\\dx) :',exy)
exz  = 0.5*(diff(u1, z) + diff(w1, x) )
print('exz = 1\\2 gxz = 1\\2(du\\dz+dw\\dx) :',exz)
eyz  = 0.5*(diff(v1, z) + diff(w1, y) )
print('eyz = 1\\2 gxy = 1\\2(dv\\dz+dw\\dy) :',eyz)

print('Matriz das deformações, [D] :\n')
def mD(exx,exy,exz,eyy,eyz,ezz):
    return np.array([[exx,exy,exz],[exy,eyy,eyz],[exz,eyz,ezz]])

mD0 = mD(exx,exy,exz,eyy,eyz,ezz)
print(mD0)
```

resultando,

\begin{equation*}
\varepsilon_{ij} =
\begin{bmatrix}
2Ax + By & (Bx + C)/2 & 0 \\
(Bx + C)/2 & 2By & M/2 \\
0 & M/2 & 2Mz
\end{bmatrix}~\textrm{micro def}
\end{equation*}

Contudo, a operação inversa, isto é, a determinação das três componentes do vetor deslocamento num ponto a partir das seis componentes das deformações é um operação mais complexa. Nesta operação invoca-se o fato de que as seis componentes de deformação não podem ser prescritas arbitrariamente, devendo existir relações entre estes. Estas relação são conhecidas por relações de contrangimento. Existem seis equações de constrangimento, agrupadas em dois grupos.

#### Primeiro grupo (I)

Partindo de:

\begin{equation*}
\begin{matrix}
\varepsilon_{xx} = \displaystyle\frac{\partial u}{\partial x}
& \varepsilon_{yy} = \displaystyle\frac{\partial v}{\partial y}
& \varepsilon_{zz} = \displaystyle\frac{\partial w}{\partial z}
\end{matrix}
\end{equation*}

Diferenciando $\varepsilon_{xx}$ e $\varepsilon_{yy}$ tal que:

\begin{equation*}
\begin{aligned}
\frac{\partial^2 \varepsilon_{xx}}{\partial y^2} &= \frac{\partial^3 u}{\partial x\partial y^2}
= \frac{\partial^2}{\partial x\partial y} \left( \frac{\partial u}{\partial y} \right)
\\[1ex]
\frac{\partial^2 \varepsilon_{yy}}{\partial x^2} &= \frac{\partial^3 v}{\partial y\partial x^2}
= \frac{\partial^2}{\partial x\partial y} \left( \frac{\partial v}{\partial x} \right)
\end{aligned}
\end{equation*}

adicionando as duas expressões resulta,

\begin{equation*}
\begin{aligned}
\frac{\partial^2}{\partial x\partial y} \left( \frac{\partial u}{\partial y} + \frac{\partial v}{\partial x} \right)
&= \frac{\partial^2 \gamma_{xy}}{\partial x\partial y} \\[1ex]
\mathrm{I.1}:~\frac{\partial^2 \varepsilon_{xx}}{\partial y^2} +
\frac{\partial^2 \varepsilon_{yy}}{\partial x^2}
&= \frac{\partial^2 \gamma_{xy}}{\partial x\partial y}
\end{aligned}
\end{equation*}

De forma análoga se obtem:

\begin{equation*}
\begin{aligned}
\mathrm{I.2}:~\frac{\partial^2 \varepsilon_{yy}}{\partial z^2} +
\frac{\partial^2 \varepsilon_{zz}}{\partial y^2}
&= \frac{\partial^2 \gamma_{yz}}{\partial y\partial z} \\[1ex]
\mathrm{I.3}:~\frac{\partial^2 \varepsilon_{xx}}{\partial z^2} +
\frac{\partial^2 \varepsilon_{zz}}{\partial x^2}
&= \frac{\partial^2 \gamma_{xz}}{\partial x\partial z}
\end{aligned}
\end{equation*}

#### Segundo grupo (II)

Este grupo estabele as relações entre as deformações de corte. Partindo de:

\begin{equation*}
\begin{matrix}
\gamma_{xy} = \displaystyle\frac{\partial u}{\partial y} + \frac{\partial v}{\partial x}
& \gamma_{xy} = \displaystyle\frac{\partial v}{\partial z} + \frac{\partial w}{\partial y}
& \gamma_{xy} = \displaystyle\frac{\partial u}{\partial z} + \frac{\partial w}{\partial x}
\end{matrix}
\end{equation*}

Diferenciando,

\begin{equation*}
\begin{aligned}
\frac{\partial \gamma_{xy}}{\partial z} &= \frac{\partial^2 u}{\partial y\partial z} +
\frac{\partial^2 v}{\partial x\partial z} \\[1ex]
\frac{\partial \gamma_{yz}}{\partial x} &= \frac{\partial^2 v}{\partial x\partial z} +
\frac{\partial^2 w}{\partial x\partial y} \\[1ex]
\frac{\partial \gamma_{xz}}{\partial y} &= \frac{\partial^2 u}{\partial y\partial z} +
\frac{\partial^2 w}{\partial x\partial y}
\end{aligned}
\end{equation*}

Adicionando as duas primeiras equações e subtraindo a terceira resulta:

\begin{equation*}
\begin{aligned}
\frac{\partial \gamma_{xy}}{\partial z}
+ \frac{\partial \gamma_{yz}}{\partial x}
- \frac{\partial \gamma_{xz}}{\partial y}
= 2 \frac{\partial^2 w}{\partial x\partial y}
\end{aligned}
\end{equation*}

Diferenciando uma vez mais esta última equação em ordem a $z$ e notando que:

\begin{equation*}
\begin{aligned}
\frac{\partial^3 w}{\partial x\partial y\partial z}
= \frac{\partial^2 \varepsilon_{zz}}{\partial x\partial y}
\end{aligned}
\end{equation*}

vem finalmente,

\begin{equation*}
\mathrm{II.1}:\frac{\partial}{\partial z}\left(\frac{\partial \gamma_{yz}}{\partial x}
+ \frac{\partial \gamma_{xz}}{\partial y}
- \frac{\partial \gamma_{xy}}{\partial z}\right)  =
2\frac{\partial^3 w}{\partial x\partial y\partial z} =
2\frac{\partial^2 \varepsilon_{zz}}{\partial x\partial y}
\end{equation*}

De forma analoga se obtem:

\begin{equation*}
\begin{aligned}
\mathrm{II.2}:\frac{\partial}{\partial x}\left(\frac{\partial \gamma_{yz}}{\partial y}
+ \frac{\partial \gamma_{xy}}{\partial x}
- \frac{\partial \gamma_{yz}}{\partial x}\right)  &=
2\frac{\partial^2 \varepsilon_{xx}}{\partial y\partial z} \\[1ex]
\mathrm{II.3}:\frac{\partial}{\partial y}\left(\frac{\partial \gamma_{xy}}{\partial z}
+ \frac{\partial \gamma_{yz}}{\partial x}
- \frac{\partial \gamma_{xz}}{\partial y}\right)
&=  2\frac{\partial^2 \varepsilon_{yy}}{\partial x\partial z}
\end{aligned}
\end{equation*}

É possível dar uma interpretação geométrica às seis equações acima obtidas. Para esse propósito, considere-se um corpo elástico discretizado em pequenos paralelepípedos, submetido a um estado de deformação definido pelas seis componentes do tensor das deformações num determinado sistema de coordenadas. É pausível conceber que, se as componentes de deformação não estiverem relacionadas entre si por certas relações, é impossível garantir uma configuração deformada contínua do sólido (haveria zonas de fendas ou interpenetração do material). Desta forma, as relações de compatibilidade de Saint-Venant fornecem essas condições. Por isso, estas são também conhecidas por equações de continuidade.


As componentes cartesianas do tensor das deformações infinitésimais devem satisfazer equações de compatibilidade assim definidas:

\begin{equation*}
\begin{aligned}
\frac{\partial^2 \varepsilon_{xx}}{\partial y^2} + \frac{\partial^2 \varepsilon_{yy}}{\partial x^2}  &=
2\frac{\partial^2 \varepsilon_{xy}}{\partial x\partial y} = \frac{\partial^2 \gamma_{xy}}{\partial x\partial y}\\
\frac{\partial^2 \varepsilon_{yy}}{\partial z^2} + \frac{\partial^2 \varepsilon_{zz}}{\partial y^2}  &=
2\frac{\partial^2 \varepsilon_{yz}}{\partial y\partial z} = \frac{\partial^2 \gamma_{yz}}{\partial y\partial z}\\
\frac{\partial^2 \varepsilon_{zz}}{\partial x^2} + \frac{\partial^2 \varepsilon_{xx}}{\partial z^2}  &=
2\frac{\partial^2 \varepsilon_{zx}}{\partial z\partial x} = \frac{\partial^2 \gamma_{zx}}{\partial z\partial x}\\
\frac{\partial}{\partial x}\left(-\frac{\partial \varepsilon_{yz}}{\partial x} + \frac{\partial \varepsilon_{zx}}{\partial y} + \frac{\partial \varepsilon_{xy}}{\partial z}\right)   &=
\frac{\partial^2 \varepsilon_{xx}}{\partial y\partial z} \\
\frac{\partial}{\partial y}\left(\frac{\partial \varepsilon_{yz}}{\partial x} - \frac{\partial \varepsilon_{zx}}{\partial y} + \frac{\partial \varepsilon_{xy}}{\partial z}\right)   &=
\frac{\partial^2 \varepsilon_{yy}}{\partial z\partial x} \\
\frac{\partial}{\partial z}\left(\frac{\partial \varepsilon_{yz}}{\partial x} + \frac{\partial \varepsilon_{zx}}{\partial y} - \frac{\partial \varepsilon_{xy}}{\partial z}\right)   &=
\frac{\partial^2 \varepsilon_{zz}}{\partial x\partial y} \\
\end{aligned}
\end{equation*}

```python
print('Exact analytical derivatives and numerical derivatives from finite differences are computed in Python with Sympy '
      '(Symbolic Python) and the Scipy.misc derivative function.')
# componentes cartesianas da matriz das deformações

Ec11m = Derivative(exx, y, y).doit() + Derivative(eyy, x, x).doit()
Ec12m = 2*Derivative(exy, x, y).doit()
print('\nEquação compatibilidade 1 ? > ',Ec11m,' = ',Ec12m)

Ec21m = Derivative(eyy, z, z).doit() + Derivative(ezz, y, y).doit()
Ec22m = 2*Derivative(eyz, y, z).doit()
print('\nEquação compatibilidade 2 ? > ',Ec21m,' = ',Ec22m)

Ec31m = diff(diff(ezz, x),x) + diff(diff(exx, z),z)
Ec32m = 2*diff(diff(exz, x),z)
print('\nEquação compatibilidade 3 ? > ',Ec31m,' = ',Ec32m)

Ec41m = diff( -diff(eyz, x) + diff(exz, y)  + diff(exy, z), x)
Ec42m = diff(diff(exx, y),z)
print('\nEquação compatibilidade 4 ? > ',Ec41m,' = ',Ec42m)

Ec51m = diff(diff(eyz, x) - diff(exz, y)  + diff(exy, z), y)
Ec52m = diff(diff(eyy, x),z)
print('\nEquação compatibilidade 5 ? > ',Ec51m,' = ',Ec52m)

Ec61m = diff(diff(eyz, x) + diff(exz, y)  - diff(exy, z), z)
Ec62m = diff(diff(ezz, x),y)
print('\nEquação compatibilidade 6 ? > ',Ec61m,' = ',Ec62m)
```
