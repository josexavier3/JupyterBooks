## Resolução

A direção do eixo dos $zz$ é direção principal de tensão. É enunciado um estado plano de tensão no plano $x-y$ tal que
a matriz das tensões é:

\begin{equation*}
\underline{\underline{\sigma}}=
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} \\
\tau_{xy} & \sigma_{yy}
\end{bmatrix} =
\begin{bmatrix}
\sigma_{xx} & 40 \\
40 & 75
\end{bmatrix}~\textrm{(MPa)}
\end{equation*}

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au03/Au03_P006_i2.jpg"
width="350" height="240">

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au03/Au03_P006_i3.jpg"
width="350" height="310">

Por definiçáo, a tensão de corte máxima é   :

\begin{equation*}
\tau_\mathrm{max} = \frac{\sigma_\mathrm{I}-\sigma_\mathrm{III}}{2}
= \frac{\sigma_\mathrm{I}}{2} \leq 50~\textrm{MPa}
\end{equation*}

Tensão principal $\sigma_\mathrm{I}$ é determinada pela seguinte equação
em função das componentes cartesianas do tensor das tensões:

\begin{equation*}
\sigma_\mathrm{I} = \frac{\sigma_{xx}+\sigma_{yy}}{2} +
\sqrt{\left(\frac{\sigma_{xx}-\sigma_{yy}}{2}\right)^2 + \tau_{xy}^2} \leq 100~\textrm{MPa}
\end{equation*}

desenvolvendo algebricamente,

\begin{equation*}
\begin{aligned}
&\sigma_\mathrm{I} - \frac{\sigma_{xx}+\sigma_{yy}}{2} = \sqrt{\left(\frac{\sigma_{xx}-\sigma_{yy}}{2}\right)^2 + \tau_{xy}^2} \\
&\left( \sigma_\mathrm{I} - \frac{\sigma_{xx}+\sigma_{yy}}{2} \right)^2 = \left(\frac{\sigma_{xx}-\sigma_{yy}}{2}\right)^2 + \tau_{xy}^2 \\
& \sigma_\mathrm{I}^2 - \sigma_\mathrm{I}(\sigma_{xx} + \sigma_{yy}) +
(\sigma_{xx}^2 + 2 \sigma_{xx}\sigma_{yy} + \sigma_{yy}^2)/4  =
(\sigma_{xx}^2 - 2 \sigma_{xx}\sigma_{yy} + \sigma_{yy}^2)/4 + \tau_{xy}^2 \\
& \sigma_\mathrm{I}^2 - \sigma_\mathrm{I}\sigma_{xx} - \sigma_\mathrm{I}\sigma_{yy} + \sigma_{xx}\sigma_{yy} = \tau_{xy}^2\\
&\sigma_{xx} = \frac{\tau_{xy}^2+\sigma_\mathrm{I}(\sigma_{yy}-\sigma_\mathrm{I})}{\sigma_{yy}-\sigma_\mathrm{I}}
\end{aligned}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import solve
from sympy import symbols

print('code 1: sympy.solvers---------------------')

sXX = symbols('sXX')

def tensaoP1(sXX,sYY,sXY):
    return (sXX + sYY)/2 + sy.sqrt((sXX - sYY)**2/4 + sXY**2)
tmax = 50 # unit: MPa
sI = 2*tmax

Syy = 75 # unit: MPa
Sxy = 40

sol = solve([tensaoP1(sXX,Syy,Sxy) - sI],sXX,dict=True)
resSxx = sol[0][sXX]
print('Sxx (\leq)',resSxx,'MPa')

print('code 2: expressão direta-------------------')
resSxx3 = (Sxy**2 + sI*(Syy-sI))/(Syy-sI)
print('Sxx (\leq)',resSxx3,'MPa')
```
