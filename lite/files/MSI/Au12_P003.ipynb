## Problema 3 <a name="PV123"></a>

O veio maciço $AB$ com 40 mm de diâmetro e 2 m de comprimento representado na figura é constituído por um material que se considera ter comportamento elastoplástico perfeito ($\tau_C$ = 145 MPa e $G$ = 77 GPa ). Na extremidade $B$ do veio é aplicado um momento de torção $T = 2125$ N.m e em seguida esse momento é removido.

[a)](#A) Determine a distribuição de tensões tangenciais numa secção transversal do veio quando o momento $T$ é aplicado.

[b)](#B) Obtenha a distribuição de tensões residuais e a rotação residual da extremidade $B$ quando $T$ é removido.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P003_i1.jpg"
width="400" height="200">

```python
import numpy as np

# Dados
# unidades: m, Pa, N

L, d = 2., 40.e-3 # m
r = d/2
TauC, G = 145.e6, 77.e9 # Pa

T = 2125. # N.m
```

#### a) Determine a distribuição de tensões tangenciais numa secção transversal do veio quando o momento $T$ é aplicado.<a name="A"></a> ([_enunciatum_](#PV123))

**Questão**

O que ocorre no veio pela aplicação do momento torsor $T$?

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P003_i5.jpg"
width="900">

Para o momento torsor aplicado é necessário averiguar, antes de mais, qual dos três cenário ocorre:

- (1) distorção angular do veio no regime elástico;
- (2) rotação parcialmente plástica, com formação de um núcleo elástico ($0<\rho<\rho_C$) e formação de uma coroa plástica ($\rho_C<\rho<c$);
- (3) rotação com plasticidade completa do veio ($\rho_C \rightarrow 0$).

**Hipótese**

Considere-se por hipótese que a resposta à questão anterior é dada pela distribuição de tensões tangenciais na secção transversal do veio definida na figura. Comente esta solução.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au12/Au12_P003_i2.jpg"
width="400">

**Verificação**

- $T < T_C$ ?  (veio elástico)

Para um veio maciço o momento elástico máximo é determinado por:

\begin{equation*}
T_C = \frac{\tau_C J}{c}~,\quad J = \frac{\pi}{2}c^4\quad\therefore\quad
T_C = \frac{\pi}{2}\tau_Cc^3
\end{equation*}

```python
J = np.pi/2*r**4
print(f'J = {J:.3e} m3')
Tc = TauC*J/r
print(f'Tc = {Tc:.2f} N.m')
```

- $T \rightarrow T_P$ ? (veio completamente plástico)

Para veios maciços, o momento máximo ou momento de torção plástico é dado pelo seguinte desenvolvimento algébrico:

\begin{equation*}
\begin{aligned}
T &= 2\pi \int_0^{\rho_C} \rho^2\left(\frac{\tau_C}{\rho_C}\rho\right)\mathrm{d}\rho +
2\pi\int_{\rho_C}^C \rho^2\tau_C\mathrm{d}\rho \\
&= \frac{2}{3}\pi c^3\tau_C\left(1-\frac{1}{4}\frac{\rho_C^3}{c^3}\right) \\
&= \frac{4}{3}T_C\left(1-\frac{1}{4}\frac{\rho_C^3}{c^3}\right) \\
&= \frac{4}{3}T_C\left(1-\frac{1}{4}\frac{\phi_C^3}{\phi^3}\right)
\end{aligned}
\end{equation*}

No limite quando a região plástica expande para a área de toda a secção, i.e., $\rho_C \rightarrow 0$:

\begin{equation*}
T_P = \frac{4}{3}T_C
\end{equation*}

```python
Tp = 4/3*Tc
print(f'Tp = {Tp:.2f} N.m')
```

Pelo que se conclui neste caso que o momento torsor aplicado está compreendido entre estes valores extremos:

\begin{equation*}
\therefore\quad T_C < T < T_P
\end{equation*}

ou seja, o veio está parcialmente plastificado, com formação de um núcleo elástico ($0<\rho<\rho_C$) e formação de uma coroa plástica ($\rho_C<\rho<c$).

Para a completa representação da distribuição da tensão de corte na secção ao longo da distância radial, i.e., a determinação da função $\tau(\rho)$, é necessário determinar o valor de $\rho_C$, ou seja, a distância radial que define o limite de elasticidade separando a região central elástica da região periférica plástica. Para esse cálculo, comece-se por construir o diagrama momento torsor - ângulo de torção ($T-\phi$).

- Ponto ($\phi_C,~T_C$):

O ângulo de torção correspondente ao momento de torção crítico, $T_C$, será o ângulo de torção máximo elástico, $\phi_C$; este vem expresso por:

\begin{equation*}
\phi_C = \frac{T_C L}{JG}
\end{equation*}

```python
phiC = Tc*L/J/G
print(f'phiC = {phiC:.4f} radianos | {np.rad2deg(phiC):.2f} º')
```

- Ponto ($\phi,~T$) + $\rho_C$ (distância radial do núcleo elástico):

Para veio maciços a uma vez que se verifica: $\rho_C / c = \phi_C / \phi$, resulta uma expressão para a função momento torsor $T$ e ângulo de torção, $\phi$ dado por:

\begin{equation*}
T = \frac{4}{3}T_C\left(1-\frac{1}{4}\frac{\phi_C^3}{\phi^3}\right)
\end{equation*}

Esta última expressão pode ser resolvida para o ângulo de torção $\phi$:

\begin{equation*}
\frac{\phi_C^3}{\phi^3} = 4 - \frac{3T}{T_C}
~\Leftrightarrow~
\phi = \phi_C/\left(4-\frac{3T}{T_C}\right)^{1/3}
\end{equation*}

```python
num = (4-3*T/Tc)**(1/3)
phi = phiC/num
print(f'phi = {phi:.4f} radianos | {np.rad2deg(phi):.2f} º')
```

O valor da posição radial de fronteira entre o domínio elástico (no centro) e o domínio plástico (na periferia) pode ser determinado por:

\begin{equation*}
\frac{\phi_C}{\rho_C} = \frac{\phi}{c}
~\Leftrightarrow~
\rho_C = c \frac{\phi_C}{\phi}
\end{equation*}

```python
rhoC  = phiC/phi*r
print(f'rhoC = {rhoC*1e3:.2f} mm')
```

Em alternativa, o núcleo de material do domínio elástico, circunscrito por uma coroa circular de material plastificado, pode ser obtido resolvendo para $\rho$ a seguinte expressão:

\begin{equation*}
\begin{aligned}
T &= \frac{4}{3}T_C\left(1-\frac{1}{4}\frac{\rho_C^3}{c^3}\right)
~\Leftrightarrow~
\rho_C = c\left(4-\frac{3T}{T_C}\right)^{1/3}
\end{aligned}
\end{equation*}

```python
num2 = (4-3*T/Tc)**(1/3)
rhoC2  = r*num2
print(f'rhoC (versão 2) = {rhoC2*1e3:.2f} mm')
```

```python
Area = np.pi*r**2
print(f'Area (veio circular maciço) = {Area*1e6:.2f} mm^2')

AreaElastica = np.pi*rhoC**2
print(f'Area (elástica) = {AreaElastica*1e6:.2f} mm^2 | {AreaElastica/Area*100:.2f} %')

AreaPlastica = np.pi*(r**2-rhoC**2)
print(f'Area (AreaPlastica) = {AreaPlastica*1e6:.2f} mm^2 | {AreaPlastica/Area*100:.2f} %')
```

#### b) Obtenha a distribuição de tensões residuais e a rotação residual da extremidade $B$ quando $T$ é removido.<a name="B"></a> ([_enunciatum_](#PV123))

**Ângulo de torção residual**

No diagrama momento de torção - ângulo de torção ($T-\phi$) a parte de descarga do momento aplicado será representado por um segmento de reta, com declive igual ao troço linear. De notar que o ângulo de torção não volta a zero depois da remoção do momento torsor aplicado, resultando uma deformação angular permanente obtida por:

\begin{equation*}
\begin{aligned}
\phi_R = \phi - \phi'
\quad\wedge\quad
\phi =  \phi_C/\left(4-\frac{3T}{T_C}\right)^{1/3}
\quad\wedge\quad
\phi' = \displaystyle\frac{T L }{JG}
\end{aligned}
\end{equation*}

em que,

- $\phi~:~$é o ângulo para a aplicação da deformação plástica desenvolvida;
- $\phi'~:~$ é o ângulo recuperado na fase de descarga ($TL/JG$) (porção elástica).

```python
phidescarga  = T*L/J/G
print(f'phi(descarga) = {phidescarga:.5f} rad | {np.rad2deg(phidescarga):.2f} º')

phiresidual = phi - phidescarga
print(f'phiresidual = {phiresidual:.5f} rad | {np.rad2deg(phiresidual):.2f} º')
```

**Distribuição de tensões de corte residuais**

As tensões residuais são obtidas pelo princípio de sobreposição. Para os pontos radiais no limite de elasticidade e na periferia do veio as tensões residuais são assim obtidas:

\begin{equation*}
\tau_1^R =  \tau_C + \tau_1
\quad\wedge\quad
\tau_2^R =  \tau_C + \tau_2
\end{equation*}

Na descarga, a tensão no limite elástico ($\rho = \rho_C$) e na extremidade do veio ($\rho = c$) podem ser obtidos por:

\begin{equation*}
\tau_1 = \displaystyle\frac{(-T)\rho}{J} \left|_{\rho = \rho_C}\right.
\quad\wedge\quad
\tau_2 = \displaystyle\frac{(-T)\rho}{J} |_{\rho = c}
\end{equation*}

```python
print('Tensões na descarga: ---------')
tau1 = -T*rhoC/J
print(f'tau1 = {tau1*1e-6:.2f} MPa')
tau2 = -T*r/J
print(f'tau2 = {tau2*1e-6:.2f} MPa')

print('Tensões residusias (pelo princípio da sobreposição): ---------')
tau1R = TauC + tau1
print(f'tau1 Residual = {tau1R*1e-6:.2f} MPa')

tau2R = TauC + tau2
print(f'tau2 Residual = {tau2R*1e-6:.2f} MPa')
```

De notar que as tensões residuais são positivas ou negativas em zonas distintas na secção do veio. Esta condição é necessária para se verificar a condição de que:

\begin{equation*}
T = \int \rho(\tau\mathrm{d}A) = 0
\end{equation*}
