## Problema 2 <a name="PV112"></a>

O tanque da figura suporta a pressão de 600 kPa e é formado por uma chapa com 10 mm de espessura soldada ao longo de um cordão contínuo que faz um ângulo $\beta = 20^\circ$ com o plano transversal. O tanque tem 1,5 m de diâmetro e está suportado nos pontos A e B que distam 3 m um do outro. Devido a uma cedência da base de apoio, o suporte A rodou em torno do eixo x um ângulo $\phi = + 0,1^\circ$.

[a)](#A) Determine o tensor das tensões no ponto P situado sobre o cordão de soldadura e na intersecção da superfície do tanque com o plano $xz$.

[b)](#B) Desenhe o diagrama de Mohr para esse estado de tensão.

[c)](#C) Utilizando o diagrama de Mohr, determine a tensão normal e tangencial no cordão de soldadura.
