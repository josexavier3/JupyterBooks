## Resolução
#### a) Da aresta $AB$.<a name="A"></a>

([enunciatum](#PIII68))


**Da lei de Hooke para um material isotrópico.**

Para um material isotrópico a lei de Hooke tridimensional vem expressa por:

\begin{equation*}
\varepsilon_{ij} = \frac{1}{E}[(1+\nu)\sigma_{ij} - \nu\delta_{ij}\sigma_{kk}]
\end{equation*}

Em forma matricial a lei constitutiva linear elástica expressa em termos de módulo de Young e coeficiente de Poisson em estado plano de tensão ($\sigma_{31} = \sigma_{13} = \sigma_{32} = \sigma_{23} = \sigma_{33}$) vem:

\begin{equation*}
\begin{bmatrix}
\varepsilon_{11} \\ \varepsilon_{22} \\ 2\varepsilon_{12}
\end{bmatrix} = \frac{1}{E}
\begin{bmatrix}
1 & -\nu & 0 \\
-\nu & 1 & 0 \\
0 & 0 & 2(1+\nu) \\
\end{bmatrix}
\begin{bmatrix}
\sigma_{11} \\ \sigma_{22} \\ \tau_{12}
\end{bmatrix}
\end{equation*}

com ainda,

\begin{equation*}
\varepsilon_{33} = -\frac{\nu}{E}(\sigma_{11} + \sigma_{22})
\end{equation*}

Em notação tensorial vem:

\begin{equation*}
\begin{bmatrix}
\varepsilon_{11} & \varepsilon_{12} \\
\varepsilon_{12} &  \varepsilon_{22}
\end{bmatrix} =
\frac{1+\nu}{E}
\begin{bmatrix}
\sigma_{11} & \tau_{12} \\
\tau_{12} &  \sigma_{22}
\end{bmatrix} -
\frac{\nu}{E}[I](\sigma_{11} + \sigma_{22})
\end{equation*}

Conhecido o estado de deformação é possível determinar a variação de comprimento de uma aresta considerando a deformação linear, e.g.:

\begin{equation*}
\begin{aligned}
\Delta L_\mathrm{AB} &= \varepsilon_{xx} L_\mathrm{AB}\quad\mathrm{(mm)} \\
\Delta L_\mathrm{BC} &= \varepsilon_{yy} L_\mathrm{AB}\quad\mathrm{(mm)} \\
\Delta t &= \varepsilon_{zz} t \quad\quad\,\,\,\mathrm{(mm)} \\
\end{aligned}
\end{equation*}

```python
import numpy as np
from matplotlib import rcParams, rc
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 18
rc('text', usetex=True)

E = 200*1e3 # unit: MPa
v = 0.3
G = E/2/(1+v)
print('\nPropriedades do material: ---------------')
print(f'     E (MPa) = {E}')
print(f'           v = {v}')
print(f'     G (MPa) = {G:.2f}')

def S2D(E,v):
    return (1/E)*np.array([[1,-v,0],[-v,1,0],[0,0,2*(1+v)]])

def C2D(E,v):
    return (E/(1-v**2))*np.array([[1,v,0],[v,1,0],[0,0,(1-v)/2]])

def tens2D(s11,s22,s12):
    return np.array([s11,s22,s12])

def def2D(e11,e22,e12):
    return np.array([e11,e22,2*e12])

sXX, sYY = 250, 200 # unit: MPa
print('\nEstado de tensão no plano x-y:')
print(f'sXX  = {sXX} MPa')
print(f'sYY  = {sYY} MPa')

print('\nEstado de deformação:')
eps = np.dot(S2D(E,v),tens2D(sXX,sYY,0))
print('Eps  = ',eps)

eps1 = (1+v)/E*np.array([[sXX,0],[0,sYY]]) - (v/E)*np.eye(2)*(sXX + sYY)
print('eps1 = \n',eps1)

print('\nComponentes da matriz das deformações:')
epsX = eps[0]
print(f'EpsX = {epsX*1e6:.1f} micro-def')
epsY = eps[1]
print('EpsY = {epsY*1e6:.1f} micro-def')
epsZ = -(v/E)*(sXX+sYY)
print('EpsZ = {epsZ*1e6:.1f} micro-def')

LAB = 300 # unit: mm
print('\n     Comprimento inicial LAB = ',LAB,' mm')
DLAB = epsX*LAB
print('Variação de comprimento DLAB = ',DLAB,' mm')
```

#### b) Da aresta $BC$.<a name="B"></a>

([enunciatum](#PIII68))

Por definição de deformação linear infinitesimal:

\begin{equation*}
\Delta L_\mathrm{BC} = \varepsilon_{yy} L_\mathrm{BC}\quad\mathrm{(mm)}
\end{equation*}

```python
LBC = 200 # unit: mm
print('     Comprimento inicial LBC = ',LBC,' mm')
DLBC = epsY*LBC
print('Variação de comprimento DLBC = ',DLBC,' mm')
```

#### c) Do volume.<a name="C"></a>

([enunciatum](#PIII68))

A variação de volume pode ser obtida por:

\begin{equation*}
\Delta V = V_f - V_0
\end{equation*}

com,

\begin{equation*}
\begin{aligned}
V_0 &= L_x \times L_y \times L_z\\
V_f &= L_x(1+\varepsilon_{xx}) \times L_y(1+\varepsilon_{yy}) \times L_z(1+\varepsilon_{zz})
\end{aligned}
\end{equation*}

Por outro lado, a extensão volumétrica pode ser obtida em função das componentes do tensor das deformações:

\begin{equation*}
\begin{aligned}
(\varepsilon_\mathrm{V} +1 )^2 &= \det(\underline{\underline{I}} + 2\underline{\underline{E}}) \quad \Longrightarrow\\
\varepsilon_\mathrm{V}^2 + 2\varepsilon_\mathrm{V} + 1 &= 
1 + 2(E_{11}+E_{22}+E_{33}) + \textrm{produtos de $E_{ij}$}
\quad \Longrightarrow
\varepsilon_\mathrm{V} &= E_{11}+E_{22}+E_{33} = E_{ii} = \mathrm{tr}(\underline{\underline{E}})
\end{aligned}
\end{equation*}

A extensão volumétrica vem assim:

\begin{equation*}
\begin{aligned}
\varepsilon_\mathrm{V} = \frac{\Delta V}{V}
= \mathrm{tr}(\underline{\underline{E}})
\approx \mathrm{tr}(\underline{\underline{D}})
= \varepsilon_{xx} + \varepsilon_{yy} + \varepsilon_{zz}
\end{aligned}
\end{equation*}

Por outro lado,

\begin{equation*}
\begin{aligned}
\varepsilon_\mathrm{V} &=  \frac{\Delta V}{V}
= \det(\underline{\underline{F}}) - 1
= \det(\underline{\underline{I}}+\underline{\underline{H}}) - 1 \\
 &= \sqrt{\det(\underline{\underline{I}}+2\underline{\underline{E}})} - 1
 \approx \sqrt{\det(\underline{\underline{I}}+2\underline{\underline{D}})} - 1
\end{aligned}
\end{equation*}

de onde,

\begin{equation*}
\begin{aligned}
\Delta V \approx \left(\sqrt{\det(\underline{\underline{I}}
+ 2\underline{\underline{D}})} - 1 \right)V
\end{aligned}
\end{equation*}

```python
print('==================================')
print('Versão 1:')
print('==================================')
t = 10 # espessura - unit : mm
V0 = LAB*LBC*t
print(f'V0 = {V0:.1f} mmm^3')

Vdef = LAB*(1+epsX) * LBC*(1+epsY) * t*(1+epsZ)
print(f'Vdef = {Vdef:.1f} mmm^3')

DV1 = Vdef - V0
print(f'Variação de volume = {DV1:.2f} mm^3')

print('==================================')
print('Versão 2:')
print('==================================')
tD0 = np.array([[epsX,0,0],[0,epsY,0],[0,0,epsZ]])
print(' tD0 = \n',tD0)
eV = np.trace(tD0)
print(f'        eV = tr(e) = {eV*1e6:.1f} micro-def')

DV2 = eV*V0
print(f'Variação de volume = {DV2:.1f} mm^3')

print('==================================')
print('Versão 3:')
print('==================================')
eV2 = np.sqrt(np.linalg.det(np.eye(3) + 2*tD0))-1
DV3 = eV2*V0
print(f'Variação de volume - [det(I+2D)])^(1\\2)-1 = {DV3:.2f} mm^3')
```
