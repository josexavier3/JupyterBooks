## Resolução

#### a) A tensão normal nas barras BE e CD.<a name="A"></a> ([enunciatum](#PIV85))

Conhecida a geometria ($L,A$) e o material ($E$) das barras do mecanismo, a tensão normal ($\sigma=F/A$) que atua nas barras $\mathrm{BE}$ e $\mathrm{CD}$ pode ser obtida em função das forças axiais que nelas atuam, respetivamente:

\begin{equation*}
\begin{cases}
F_\mathrm{BE} &= \delta_\mathrm{B}\displaystyle\frac{AE}{L_\mathrm{BE}}
\quad\wedge\quad \sigma_\mathrm{BE}=\frac{F_\mathrm{BE}}{A}\\
F_\mathrm{CD} &= \delta_\mathrm{C}\displaystyle\frac{AE}{L_\mathrm{CD}}
\quad\wedge\quad \sigma_\mathrm{BE}=\frac{F_\mathrm{CD}}{A}
\end{cases}
\end{equation*}

A solução destas equações é direta uma vez conhecidas as variações de comprimento $\delta_\mathrm{B}$ e $\delta_\mathrm{C}$. Para tal considere-se a análise do equilíbrio e deformada da barra $\mathrm{ABC}$, bem como as equações de compatibilidade na ligação roscada no ponto $\mathrm{B}$.

##### (i) Diagrama de corpo livre da barra $\mathrm{ABC}$

Considere-se o diagrama de corpo livre (DCL) da barra rígida $\mathrm{ABC}$, com a representação das forças externas que nela atuam. O apoio em $\mathrm{A}$ é considerado uma rótula plana bloqueando apenas as translações horizontal e vertical no apoio. As ligações roscadas nos pontos $\mathrm{B}$ e $\mathrm{C}$ irão dar origem a tensões normais de tração as barras $\mathrm{BE}$ e $\mathrm{CD}$, pelo que o sentido indicado para essas forças no DCL é compressivo.

(De notar que o avanço da rosca em B irá rodar a barra rigida ABC em torno do ponto A; contudo localmente este movimento é traduzido por um avanço axial da barra ao longo do eixo BE. Por sua vez, a barra rigida irá exercer uma força oposta - que será de tração - na ligação e portanto no parafuso em B. Em relação ao ponto C, a rotação da barra rigida ABC irá gerar uma força de tração na barra CD. No DCL estas forças representam-se com o sentido oposto ao descrito.)

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au08/Au08_P005_i2.jpg"
width="350" height="250">

A análise do equilíbrio estático permite escrever o seguinte sistema de equações:

\begin{equation*}
\begin{cases}
+/\rightarrow~&:~\sum F_x=0
\quad\Leftrightarrow\quad
R_\mathrm{Ax} + F_\mathrm{BE} - F_\mathrm{CD}  =0 \\
+/\uparrow~&:~\sum F_y=0
\quad\Leftrightarrow\quad
R_\mathrm{Ay} = 0 \\
+/\curvearrowleft~&:~\sum M_A=0
\quad\Leftrightarrow\quad
-L_\mathrm{AC} F_\mathrm{CD} + L_\mathrm{AB} F_\mathrm{BE} =0 \\
\end{cases}
\end{equation*}

Como se pode concluir o sistema de equações não tem solução única (3 incógnitas para 2 equações), pelo que é um sistema estaticamente indeterminado. Contudo, da última equação pode-se concluir a seguinte relação entre as forças axiais que atuam nas barras $BE$ e $DC$ ($L_\mathrm{AC}=250$ mm, $L_\mathrm{AB}=150$ mm):

\begin{equation*}
\frac{F_\mathrm{CD}}{F_\mathrm{BE}} =
\frac{L_\mathrm{AB}}{L_\mathrm{AC}} = 0.6
\end{equation*}

##### (ii) Deformada (movimento de corpo rígido) da barra rígida $ABC$

Considere-se a deformada da barra rígida (não deformável) $ABC$ como ilustrado:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au08/Au08_P005_i3.jpg"
width="300" height="400">

Da análise da geometria da deformada da barra $ABC$ conclui-se, por igualdade de triângulos ($L_\mathrm{AC}=250$ mm, $L_\mathrm{AB}=150$ mm):

\begin{equation*}
\tan\theta = \frac{\delta_\mathrm{B}}{L_\mathrm{AB}} =
\frac{\delta_\mathrm{C}}{L_\mathrm{AC}}
\quad\Leftrightarrow\quad
\frac{\delta_\mathrm{B}}{\delta_\mathrm{C}} =
\frac{L_\mathrm{AB}}{L_\mathrm{AC}} = 0.6
\end{equation*}

A equação anterior estabelece uma relação entre as variações de comprimento longitudinais das barras $\mathrm{BE}$ e $\mathrm{DC}$. Para a solução do problema é necessário definir uma segunda relação. Esta é obtida pela equação de compatibilidade de deslocamentos na barra entre as secções $\mathrm{B}$ ($\delta_B$) e $\mathrm{C}$ ($\delta_C$).

##### (iii) Deslocamento relativo da ligação roscada em $B$

Pelo aperto gerado na ligação roscada no avanço da porta, $\Delta\equiv \delta_{B/p}$, o ponto $\mathrm{B}$ da barra rígida $\mathrm{ABC}$ irá sofrer um deslocamento $\delta_\mathrm{B}$. De notar que $|\Delta| > |\delta_\mathrm{B}|$, uma vez que a barra $\mathrm{BE}$ (de extremidade roscada) tem rigidez finita. Na verdade, o avanço da porca gera na barra $\mathrm{BE}$ a uma força axial de tração, que é responsável por um deslocamento $\delta_\mathrm{p}$. A equação de compatibilidade entre estes deslocamento é dada por:

\begin{equation*}
\Delta - \delta_{p} = \delta_{B}~,\quad
\delta_{p} = \frac{F_\mathrm{BE} L_\mathrm{BE}}{AE}
\end{equation*}

##### Sistema de equações

Das (i) relações de equilíbro estático, (ii) análise da geometria do movimento de corpo rígido da placa $ABC$, (iii) das relações de deslocamento relativo gerados na ligação aparafusa, resulta o seguinte sistema de equações:

\begin{equation*}
\begin{cases}
(i)~:~F_\mathrm{CD} &= \left(\displaystyle\frac{L_\mathrm{AB}}{L_\mathrm{AC
}}\right)~F_\mathrm{BE} \\
(ii)~:~\delta_\mathrm{B} &= \left
(\displaystyle\frac{L_\mathrm{AB}}{L_\mathrm{AC}}\right)~\delta_\mathrm{C}
\quad\wedge\quad \delta_{C} = \displaystyle\frac{F_\mathrm{CD}L_\mathrm{CD}}{A_\mathrm
{CD}E} \\
(iii)~:~\delta_{B} &= \Delta - \delta_{p}~\quad\wedge\quad
\delta_{p} = \displaystyle\frac{F_\mathrm{BE}L_\mathrm{BE}}{A_\mathrm{BE}E}
\end{cases}
\end{equation*}

\begin{equation*}
\quad\Leftrightarrow\quad
\begin{cases}
(i)~&:~F_\mathrm{CD} = \left(\displaystyle\frac{L_\mathrm{AB}}{L_\mathrm{AC
}}\right)~F_\mathrm{BE}\\
(ii)\rightarrow (iii)~&:~\left(\displaystyle\frac{L_\mathrm{AB}}{L_\mathrm{AC
}}\right)
~\displaystyle\frac{F_\mathrm{CD} L_\mathrm{CD}}{AE} &=
\Delta -
\displaystyle\frac{F_\mathrm{BE} L_\mathrm{BE}}{AE}
\end{cases}
\end{equation*}


Cuja solução algébrica permite determinar:

\begin{equation*}
\left[\left(\displaystyle\frac{L_\mathrm{AB}}{L_\mathrm{AC}}\right)^2 L_\mathrm{CD}
+ L_\mathrm{BE} \right] F_\mathrm{BE} = \Delta AE
\quad\Leftrightarrow\quad
F_\mathrm{BE} = \frac{\Delta AE L_\mathrm{AC}^2}
{L_\mathrm{AB}^2L_\mathrm{CD} + L_\mathrm{BE}L_\mathrm{AC}^2}
\end{equation*}

```python
import numpy as np

# unidades: mm, N, MPa
LAB, LBC, LCD, LBE = 150, 100, 1800, 2300
LAC = LAB+LBC

dia = 16 # unidade: mm

E = 200e3 # unidade: MPa
passo = 2.5 # unidade: mm
Delta = passo

ABE = np.pi*dia**2/4 # =ADC, unidade ; mm^2
A = ABE
print(f'Area (BE\\CD) = {A:.3f} mm^2')

FBE = Delta*A*E*LAC**2/(LAB**2*LCD + LBE*LAC**2)
print(f'FBE = {FBE:.2f} N')
FCD = (LAB/LAC)*FBE
print(f'FCD = {FCD:.2f} N')
```

\begin{equation*}
\begin{aligned}
\sigma_{\mathrm{BE}} &= \frac{F_\mathrm{BE}}{A} \quad\textrm{(tração)}\\
\sigma_{\mathrm{CD}} &= \frac{F_\mathrm{CD}}{A} \quad\textrm{(tração)}
\end{aligned}
\end{equation*}

```python
sigBE = FBE/A
print(f'sigBE = {sigBE:.2f} MPa')

sigCD = FCD/A
print(f'sigCD = {sigCD:.2f} MPa')
```
