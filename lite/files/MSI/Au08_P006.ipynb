## Resolução

#### a) A tensão em cada fio provocada pela aplicação da carga.<a name="A"></a>

([enunciatum](#PIV86))

##### Diagrama de corpo livre da barra rígida $ABCD$

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au08/Au08_P006_i2.jpg"
width="350" height="140">

A análise do equilíbrio estático permite escrever o seguinte sistema de equações:

\begin{equation*}
\begin{cases}
+/\uparrow~&:~\sum F_y=0
\quad\Leftrightarrow\quad
F_A + F_C + F_D - P = 0 \\
+/\curvearrowleft~&:~\sum M_A=0
\quad\Leftrightarrow\quad
-L_\mathrm{AB}P + L_\mathrm{AC} F_C + L_\mathrm{AD} F_D =0 \\
\end{cases}
\end{equation*}

Conclui-se que o sistema é estaticamente indeterminado uma vez que as equações não têm solução única (3 incógnitas para 2 eqs.).

##### Geometria da deformada

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSI/Notebooks/Au08/Au08_P006_i3.jpg"
width="350" height="200">

\begin{equation*}
\frac{\delta_A-\delta_C}{L_\mathrm{AC}}
= \frac{\delta_C-\delta_D}{L_\mathrm{CD}}
= \frac{\delta_A-\delta_D}{L_\mathrm{AD}}
\end{equation*}

Por outro lado,

\begin{equation*}
\delta_\alpha = \frac{F_\alpha L_\alpha}{AE}
\quad\alpha=A,C,D
\end{equation*}

pelo que,

\begin{equation*}
\frac{F_A-F_C}{L_\mathrm{AC}}
= \frac{F_C-F_D}{L_\mathrm{CD}}
= \frac{F_A-F_D}{L_\mathrm{AD}}
\end{equation*}

A solução do problema consiste portanto na solução do seguinte sistema de equações, por exemplo:

\begin{equation*}
\begin{cases}
F_A + F_C + F_D - P = 0 \\
-L_\mathrm{AB}P + L_\mathrm{AC} F_C + L_\mathrm{AD} F_D =0 \\
F_A-F_C = \displaystyle\frac{L_\mathrm{AC}}{L_\mathrm{CD}}(F_C-F_D)
\end{cases}
\end{equation*}

```python
import numpy as np
from matplotlib import rcParams, rc
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
rc('text', usetex=True)

from sympy.solvers import solve
from sympy import symbols

# unidades: mm, N, MPa

FA, FC, FD = symbols('FA FC FD')

P = 5.e3 # unidade: N
E = 200.e3 # unidade: MPa
dia = 4. # unidade: mm

LAB, LBC, LCD  = 350., 350., 350.
LAC = LAB + LBC
LAD = LAC + LCD
L = 300.

A = np.pi*dia**2/4

eq1 = FA + FC + FD - P
eq2 = -LAB*P + LAC*FC + LAD*FD
eq3 = FA - FC - (LAC/LCD)*(FC - FD)

forcas = solve([eq1,eq2,eq3], [FA,FC,FD], dict=True)
FAn = forcas[0][FA]
FCn = forcas[0][FC]
FDn = forcas[0][FD]

print('forças:--------------')
print(f'FA = {FAn:.2f} N')
print(f'FC = {FCn:.2f} N')
print(f'FD = {FDn:.2f} N')

sigA = FAn/A
sigC = FCn/A
sigD = FDn/A
print('\ntensões (tração):--------------')
print(f'sigA = {sigA:.2f} MPa')
print(f'sigC = {sigC:.2f} MPa')
print(f'sigD = {sigD:.2f} MPa')
```

#### b) O deslocamento do ponto $B$.<a name="B"></a> ([enunciatum](#PIV86))

Pela geometria da deformada:

\begin{equation*}
\frac{\delta_A-\delta_B}{L_{AB}} = \frac{\delta_A-\delta_D}{L_{AD}}
\quad\Leftrightarrow\quad
\delta_B = \delta_A - \frac{L_{AB}}{L_{AD}} (\delta_A-\delta_D)
\end{equation*}

com,

\begin{equation*}
\delta_A = \frac{F_A L}{AE}~;
\quad
\delta_D = \frac{F_D L}{AE}
\end{equation*}

```python
print(f'L\\A\\E = {L/A/E:.4e} 1\\mm.MPa')
deltaA = FAn*L/A/E
deltaC = FCn*L/A/E
deltaD = FDn*L/A/E
print('\ntensões (tração): --------------')
print(f'deltaA = {deltaA:.4f} mm')
print(f'deltaC = {deltaC:.4f} mm')
print(f'deltaD = {deltaD:.4f} mm')

deltaB = deltaA - (LAB/LAD)*(deltaA-deltaD)
print(f'\ndeltaB = {deltaB:.4f} mm')
```
