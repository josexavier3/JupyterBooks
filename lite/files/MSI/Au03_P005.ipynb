## Resolução

A matriz das tensões no referencial cartesiano $S_0(O,x,y,z)$ é dado por:

\begin{equation*}
\underline{\underline{\sigma}}^T=
\begin{bmatrix}
\sigma_{xx} & \tau_{yx} & \tau_{zx} \\
\tau_{xy} & \sigma_{yy} & \tau_{zy} \\
\tau_{xz} & \tau_{yz} & \sigma_{zz}
\end{bmatrix} =
\begin{bmatrix}
60 & 0 & \tau_{zx} \\
0 & 100 & 0 \\
\tau_{xz} & 0 & 0
\end{bmatrix} \quad\textrm{(MPa)}
\end{equation*}

Pela análise direta da matriz das tensões pode concluir-se que o eixo dos $yy$ é uma direção principal de tensão
(uma vez que: $\tau_{yx} = \tau_{yz} = 0$) e que o valor de $\sigma_{yy} \equiv \sigma_{I} = 100$ MPa é uma tensão
principal (valor próprio da matriz $\sigma_{ji}$).

O estado bi-dimensional no plano $x-z$ vem dado por:

\begin{equation*}
\underline{\underline{\sigma}}^T =
\begin{bmatrix}
\sigma_{xx} & \tau_{zx} \\
\tau_{xz} & \sigma_{zz}
\end{bmatrix} =
\begin{bmatrix}
60 & \tau_{zx} \\
\tau_{xz} & 0
\end{bmatrix} \quad\textrm{(MPa)}
\end{equation*}

A determinação do intervalo de valores de $\tau_{xz}$ para a qual a tensão
tangencial máxima é sempre $\le$ 60 MPa, pode ser obtida pela análise da circunferência de Mohr. Na construção do
círculo de Morh, considere-se os seguintes dados: (i) $\sigma_{yy} = 100$ MPa é direção principal de
tensão; (ii) no plano $x-z$, o diâmetro do círculo é definido pelo segmento de reta unindo os pontos: $A(60,\tau_{xz})$
e $B(0,-\tau_{xz})$. Conclui-se neste caso: $\sigma_{yy} \equiv \sigma_\mathrm{I}$.

Atendendo ao plano formado pelas direções $\vec{n}_\mathrm{I}$ e $\vec{n}_\mathrm{III}$, a tensão de corte máxima
vem dada por:

\begin{equation*}
\begin{aligned}
|\tau_\mathrm{max} | & = \frac{\sigma_{\mathrm{I}}-\sigma_{\mathrm{III}}}{2} \leq 60~\textrm{MPa}
\end{aligned}
\end{equation*}

De onde se calcula $\sigma_\mathrm{III} \ge -20$ MPa.

Resulta também da análise do círculo do Morh no plano $x-z$ que a tensão média (ponto C) vem:

\begin{equation*}
\begin{aligned}
\sigma_{\mathrm{med}} \equiv C = \frac{\sigma_{xx}+\sigma_{zz}}{2} = \frac{60+0}{2}= 30~\textrm{MPa}
\end{aligned}
\end{equation*}

Conhecidos o centro $C$ e a tensão principal (mínima) $\sigma_\mathrm{III}$, da análise goemétrica do círculo de Morh conclui-se
que o raio $R$ no plano $x-z$ é:

\begin{equation*}
\begin{aligned}
R_{C_2}^{x-z} = \sigma_{\mathrm{med}} - \sigma_{\mathrm{III}} = 30 + 20 = 50~\textrm{MPa}
\end{aligned}
\end{equation*}

Assim sendo, da expressão do raio da circunferência:

\begin{equation*}
\begin{aligned}
& \quad |R_{C_2}^{x-z}|^2 = \left(\frac{\sigma_{xx}-\sigma_{zz}}{2}\right)^2 +
\tau_{xz}^2  \\
\Leftrightarrow & \quad
50^2 = 30^2 + \tau_{xz}^2
\quad \Leftrightarrow \quad \tau_{xz} = \pm 40~\textrm{MPa}
\end{aligned}
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams, rc
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
# rc('text', usetex=True)

def tt2D(s11,s22,s12): return np.array([[s11,s12],[s12,s22]])

txy0 = 40
S = tt2D(60,0,txy0)
S11 = S[0][0]
S12 = S[0][1]
S22 = S[1][1]
center = [(S11 + S22)/2.0, 0.0]
radius = np.sqrt((S11 - S22)**2/4.0 + S12**2)
Smin = center[0] - radius
Smax = center[0] + radius

print('Ponto A (','{:.1f}'.format(S11),',','{:.1f}'.format(S12),')')
print('Ponto B (','{:.1f}'.format(S22),',','{:.1f}'.format(-S12),')')

print("  Tensão normal mínima: ", np.round(Smin,6))
print("  Tensão normal máxima: ", np.round(Smax, 6))
print("  Tensão normal média : ", np.round(center[0], 6))
print("Tensão de corte mínima: ", np.round(-radius, 6))
print("Tensão de corte máxima: ", np.round(radius, 6))

circ = plt.Circle((center[0],0), radius, facecolor='tab:olive', lw=3,
edgecolor='tab:orange')
desv = 0.1*radius
plt.axis('image')
ax = plt.gca()
ax.add_artist(circ)
ax.set_xlim(Smin - .3*radius, Smax + .3*radius)
ax.set_ylim(-1.3*radius, 1.3*radius)
plt.plot([S22, S11], [S12, -S12], 'ko')
plt.plot([S22, S11], [S12, -S12], 'k')
plt.plot(Smin, 0, 'bs')
plt.plot(Smax, 0, 'bs')
plt.plot(center[0], radius, 'rv')
plt.plot(center[0], -radius, 'rv')
plt.plot(center[0], center[1], 'o', mfc='w')
plt.text(center[0]+desv, desv, 'C')
plt.text(S22 + desv, S12 + desv , 'A')
plt.text(S11 + desv, -S12 + desv, 'B')
plt.text(Smin - 0.25*radius, -0.15*radius, r'$\sigma_\mathrm{II}$')
plt.text(Smax + desv, -0.15*radius, r'$\sigma_\mathrm{I}$')
plt.text(center[0]-4*desv, 1.1*radius, r'$\tau_\mathrm{max}$')
plt.text(center[0]-4*desv, -1.1*radius, r'$\tau_\mathrm{min}$')
plt.xlabel(r"$\sigma$", size=18)
plt.ylabel(r"$\tau$", size=18)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

```python
sxx = 60
syy = 100
szz  = 0

sI = syy
print('sI = syy =',sI,' MPa')
smed = (sxx+szz)/2
print('smed =',smed,' MPa')

sIII = sI - 2*sxx
print('sIII =',sIII,' MPa')

Rc2 = smed - sIII
print('Rc2 =',Rc2,' MPa')

tzy = np.sqrt(Rc2**2 - (sxx-smed)**2)
print('tzy = +/-',tzy,' MPa')
```
