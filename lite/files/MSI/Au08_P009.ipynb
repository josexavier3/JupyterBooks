## Problema 9 <a name="PIV809"></a>

A barra rígida $ACEB$ representada na figura está ligada a uma articulação em $A$ e a duas barras de alumínio $DC$ e $FE$, cada uma com diâmetro de 25 mm. O coeficiente de dilatação térmico do alumínio é $\alpha_\mathrm{alumínio}~23\times 10^{-6}$ /ºC e o seu módulo de elasticidade vale $E_\mathrm{alumínio}$ = 70 GPa. Considerando que a barra rígida está inicialmente em posição vertical determine as tensões em cada haste e o deslocamento horizontal do ponto $B$ causado por um aumento de temperatura nas barras de alumínio, $\Delta T$ = 80º C.

width="300" height="440">

## Resolução

```python
import numpy as np

# ---------------------------
# DADOS do problema:
# unidades: m, Pa, N

# Modulo de Young
E = 70.e9
# barras verticais
LEF, LCD, LAB = 0.3, 0.6, 1.2
dia = 25.e-3 #unidade: m
Aalum = np.pi*dia**2/4
print(f'Aalum = {Aalum:.5e} m2')
alphaalum = 23.*1e-6 # unidade: /ºC
DT = 80. # unidade: ºC
# ---------------------------
```

A variação de temperatura $\Delta T$ irá introduzir uma dilatação do mecanismo, pelo que as barras $EF$ e $CD$ irão sofrer uma expansão com variação de comprimento de acordo com a relação:

$$\delta = \alpha \Delta T L$$

Uma vez que as barras são constituidas pelo mesmo material (alumínio), o coeficiente de dilatação térmico é igual, pelo que a variação de comprimento será apenas função do comprimento inicial das barras. Desta forma, verifica-se que : $\delta_{CD}^T > \delta_{EF}^T$.

Contudo, as barras $EF$ e $CD$ não se movem livremente, pois estão ligadas à barra rígida $AB$ que, por sua vez, está articulada em $A$. A barra rígida $AB$ sendo indeformável terá apenas um movimento de corpo rígido de rotação em torno do ponto $A$. Esta restrição de movimento do mecanismo irá gerar, forças axiais sobre as barras $EF$ e $CD$. De notar que a barra $EF$ irá ficar à tração e a barra $CD$ irá ficar submetida à compressão.

Estabelecendo o diagrama de corpo livre (DCL) da barra rigida AB (forças das barras EF e CD e forças de reação no ponto A):

##### Equilíbrio estático

Analisando o equilíbrio estático pela resultante dos momentos em torno do ponto A, resulta:

\begin{equation*}
+/\curvearrowleft~:~\sum M_A = 0
~\Leftrightarrow~
0.9F_{EF} - 0.3F_{CD} = 0
~\Leftrightarrow~
F_{CD} = 3 F_{EF}
\end{equation*}

##### Deformação gerada por uma variação de temperatura

\begin{equation*}
\delta_{EF}^T = \alpha \Delta T L_{EF} \quad \wedge \quad
\delta_{CD}^T = \alpha \Delta T L_{CD}
\end{equation*}

```python
deltaEF_T =  alphaalum*DT*LEF
print(f'deltaEF_T = {deltaEF_T*1e3:.3f} m')

deltaCD_T =  alphaalum*DT*LCD
print(f'deltaCD_T = {deltaCD_T*1e3:.3f} m')
```

##### Geometria da deformada da barra rígida / Equação de compatibilidade

Pela igualdade de triângulos:

\begin{equation*}
\tan\theta = \frac{\delta_{EF}}{0.9} = \frac{\delta_{CD}}{0.3}
= \frac{\delta_{B}}{1.2}
\end{equation*}

desenvolvendo,

\begin{equation*}
\begin{aligned}
\delta_{EF}
&= 3 \delta_{CD}\\
\delta_{EF}^T + \delta_{EF}^F
&= 3 (\delta_{CD}^T - \delta_{CD}^F) \\
\alpha \Delta T L_{EF} + \frac{F_{EF}L_{EF}}{A_{EF}E}
&= 3 \left(\alpha \Delta T L_{CD} -
\frac{F_{CD}L_{CD}}{A_{CD}E}\right) \\
\end{aligned}
\end{equation*}

##### Sistema de equações

Combinando as equações de equilíbrio estático com a equação de compatibilidade de deslocamentos resulta o seguinte sistema de equações:

\begin{equation*}
\begin{cases}
F_{CD} = 3 F_{EF} \\
\alpha \Delta T L_{EF} + \displaystyle\frac{F_{EF}L_{EF}}{A_{EF}E}
= 3 \left(\alpha \Delta T L_{CD} -
\displaystyle\frac{F_{CD}L_{CD}}{A_{CD}E}\right) \\
\end{cases}
~\Leftrightarrow~
\begin{cases}
F_{CD} = 3 F_{EF}~\quad\textrm{(N)} \\
F_{EF} = \displaystyle\frac{\alpha EA\Delta T(3L_{CD}- L_{EF})}{(L_{EF}+9L_{CD})}
~\quad\textrm{(N)}
\end{cases}
\end{equation*}

**Tensões nas hastes**

Calculadas as forças axiais nas hastes, as tensões (MPa) vêm dadas por:

\begin{equation*}
\sigma_{EF} = \displaystyle\frac{F_{EF}}{A} \quad\wedge\quad
\sigma_{CD} = \displaystyle\frac{F_{CD}}{A}
\end{equation*}

```python
num = alphaalum*E*Aalum*DT*(3*LCD - LEF)
den = (LEF + 9*LCD)

FEF = num/den
print(f'FEF = {FEF:.2f} N')
FCD = 3*FEF
print(f'FCD = {FCD:.2f} N')
```

```python
sigEF = FEF/Aalum
print(f'sigEF = {sigEF*1e-6:.2f} MPa')

sigCD = FCD/Aalum
print(f'sigCD (vers 1) = {sigCD*1e-6:.2f} MPa')

sigCDb = 3*sigEF
print(f'sigCD (vers 2) = {sigCDb*1e-6:.2f} MPa')
```

**Deslocamento horizontal do ponto B**

Variação relativa de comprimento da barra $EF$ é expresso, pelo princípio da sobreposição, pelo alongamento devido à variação de temperatura e pela força de tração resultante:

\begin{equation*}
\delta_{EF} = \alpha \Delta T L_{EF} + \frac{F_{EF}L_{EF}}{A_{EF}E}
\end{equation*}

Já o deslocamento em B pode ser, por sua vez, determinado por:

\begin{equation*}
\delta_{B} = \frac{1.2}{0.9}\,\delta_{EF} = 1.333 \,\delta_{EF}
\end{equation*}


```python
deltaEF_F = FEF*LEF/Aalum/E
print(f'deltaEF (F) (vers. 1) = {deltaEF_F*1e3:.3f} mm')

deltaEF_Fb = sigEF*LEF/E
print(f'deltaEF (F) (vers. 2) = {deltaEF_Fb*1e3:.3f} mm')

deltaEF = deltaEF_T + deltaEF_F
print(f'deltaEF (T+F) = {deltaEF*1e3:.3f} mm')

deltaB = deltaEF*1.2/0.9
print(f'deltaB (T+F) = {deltaB*1e3:.3f} mm')
```
