#### a) Qual será a leitura de um extensómetro colocado segundo a diagonal representada na figura ?<a name="A"></a>

([enunciatum](#PIII67))

##### Matriz das tensões.

De acordo com o enunciado, existe um estado de tensão plano em $x-y$. Existem apenas tensões normais aos eixos dos $xx$ e $yy$. Pelo que a matriz das tensões no referencial $S(O,x,y,z)$ vem expresso por:

\begin{equation*}
\underline{\underline{\sigma}} = \sigma_{ij} =
\begin{bmatrix}
\sigma_{xx} & 0 & 0 \\
0 & \sigma_{yy} & 0 \\
0 & 0 & 0 \\
\end{bmatrix} =
\begin{bmatrix}
100 & 0 & 0 \\
0 & 150 & 0 \\
0 & 0 & 0 \\
\end{bmatrix} \quad\textrm{(MPa)}
\end{equation*}

##### Da lei de Hooke para um material isotrópico.

Para um material isotrópico a lei de Hooke geral vem expressa por:

\begin{equation*}
\varepsilon_{ij} = \frac{1}{E}[(1+\nu)\sigma_{ij} - \nu\delta_{ij}\sigma_{kk}]
\end{equation*}

Em forma matricial a lei constitutiva linear elástica expressa em termos de módulo de Young e coeficiente de Poisson vem:

\begin{equation*}
\begin{cases}
\varepsilon_{11} &= \frac{1}{E}[\sigma_{11} - \nu(\sigma_{22}+\sigma_{33})] \\
\varepsilon_{22} &= \frac{1}{E}[\sigma_{22} - \nu(\sigma_{11}+\sigma_{33})] \\
\varepsilon_{33} &= \frac{1}{E}[\sigma_{33} - \nu(\sigma_{11}+\sigma_{22})] \\
2\varepsilon_{23} &= \gamma_{23} = \frac{\tau_{23}}{G}\\
2\varepsilon_{13} &= \gamma_{13} = \frac{\tau_{13}}{G}\\
2\varepsilon_{12} &= \gamma_{12} = \frac{\tau_{12}}{G}\\
\end{cases}
\end{equation*}

com,

\begin{equation*}
G = \frac{E}{2(1+\nu)}
\end{equation*}

Na forma matricial têm-se:

\begin{equation*}
\begin{bmatrix}
\varepsilon_{11} \\ \varepsilon_{22} \\ \varepsilon_{33} \\ 2\varepsilon_{23} \\ 2\varepsilon_{13} \\ 2\varepsilon_{12}
\end{bmatrix} = \frac{1}{E}
\begin{bmatrix}
1 & -\nu & -\nu & 0 & 0 & 0 \\
-\nu & 1 & -\nu & 0 & 0 & 0 \\
-\nu & -\nu & 1 & 0 & 0 & 0 \\
0 & 0 & 0 & 2(1+\nu) & 0 & 0 \\
0 & 0 & 0 & 0 & 2(1+\nu) & 0 \\
0 & 0 & 0 & 0 & 0 & 2(1+\nu) \\
\end{bmatrix}
\begin{bmatrix}
\sigma_{11} \\ \sigma_{22} \\ \sigma_{33} \\ \tau_{23} \\  \tau_{13} \\ \tau_{12}
\end{bmatrix}
\end{equation*}

Para um estado plano de tensão, no plano $1-2$, para o qual se verifica: $\sigma_{33} = \tau_{31} = \tau_{32} = 0$, a lei de Hooke escreve-se:

\begin{equation*}
\begin{cases}
\varepsilon_{11} &= \frac{1}{E}[\sigma_{11} - \nu\sigma_{22}] \\
\varepsilon_{22} &= \frac{1}{E}[\sigma_{22} - \nu\sigma_{11}] \\
2\varepsilon_{12} &= \gamma_{12} = \frac{\tau_{12}}{G}
= \frac{2(1+\nu)}{E}\tau_{12}\\
\end{cases}
\end{equation*}

com,

\begin{equation*}
\varepsilon_{33} = -\frac{\nu}{E}(\sigma_{11}+\sigma_{22}) \\
\end{equation*}

Por outro lado, em estado plano de deformação, para o qual se verifica: $\varepsilon_{33}=0$ e $\sigma_{33} = \nu(\sigma_{11}+\sigma_{22})$, a lei de Hooke pode ser simplifiada para:

\begin{equation*}
\begin{cases}
\varepsilon_{11} &= \frac{1-\nu^2}{E}(\sigma_{11} - \frac{\nu}{1-\nu}\sigma_{22}) \\
\varepsilon_{22} &= \frac{1-\nu^2}{E}(\sigma_{22} - \frac{\nu}{1-\nu}\sigma_{11}) \\
2\varepsilon_{12} &= \gamma_{12} = \frac{\tau_{12}}{G}
= \frac{2(1+\nu^2)}{E}\tau_{12}\\
\end{cases}
\end{equation*}

```python
import numpy as np
from sympy import symbols
import matplotlib.pyplot as plt
from matplotlib import rcParams, rc
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 18
rc('text', usetex=False)

np.set_printoptions(precision=10)

def mtensoes(sXX,sXY,sXZ,sYY,sYZ,sZZ):
    return np.array([[sXX,sXY,sXZ],[sXY,sYY,sYZ],[sYZ,sYZ,sZZ]])

# unit: MPa
sX, sY = 100., 150.
mtP = mtensoes(sX,0,0,sY,0,0)
print('Matriz das tensões: ---------------')
print('mtP [MPa] = \n',mtP)
```

```python
# unit: MPa
E, v = 200*1e3, 0.3
G = E/2/(1+v)
print('\nPropriedades do material: ---------------')
print('     E (MPa) = ',E)
print('           v = ',v)
print('     G (MPa) = ','{:.2f}'.format(G))

e11,e22,e33,e23,e13,e12 = symbols('e11 e22 e33 e23 e13 e12')
s11,s22,s12 = symbols('s11 s22 s12')

# estado plano de tensão

def e11(sxx,syy,szz,E): return (sxx - v*(syy+szz))/E
def e22(sxx,syy,szz,E): return (syy - v*(sxx+szz))/E
def e33(sxx,syy,szz,E): return (szz - v*(sxx+syy))/E
def e23(syz,G): return (syz/G)/2
def e13(sxz,G): return (sxz/G)/2
def e12(sxy,G): return (sxy/G)/2

print('\nComponentes da matriz das deformações infinitesimais: ---------------')
exxP = e11(mtP[0,0],mtP[1,1],mtP[2,2],E)
print('exxP = ',exxP*1e6,' micro-def')
eyyP = e22(mtP[0,0],mtP[1,1],mtP[2,2],E)
print('eyyP = ',eyyP*1e6,' micro-def')
ezzP = e33(mtP[0,0],mtP[1,1],mtP[2,2],E)
print('ezzP = ',ezzP*1e6,' micro-def')
eyzP = e23(mtP[1,2],G)
print('eyzP = ',eyzP*1e6,' micro-def')
exzP = e13(mtP[0,2],G)
print('exzP = ',exzP*1e6,' micro-def')
exyP = e12(mtP[0,2],G)
print('exyP = ',exyP*1e6,' micro-def')

def mdef(eXX,eXY,eXZ,eYY,eYZ,eZZ):
    return np.array([[eXX,eXY,eXZ],[eXY,eYY,eYZ],[eYZ,eYZ,eZZ]])

mDP = mdef(exxP,exyP,exzP,eyyP,eyzP,ezzP)
print('\nmDP = \n',mDP)
```

Para finalizar, a extensão linear numa determinada direção definida pelo ângulo $\theta$ com a direção horizontal do eixo dos $xx$ é
dada por:

\begin{equation*}
\begin{aligned}
\varepsilon_{\theta} &= \varepsilon_{xx}\cos^2\theta +
\varepsilon_{yy}\sin^2\theta + \gamma_{xy}\sin\theta\cos\theta \\
\end{aligned}
\end{equation*}

Para o problema enunciado, tem-se:

```python
def extens(exx,eyy,gxy,t):
    return exx*np.cos(t)**2 + eyy*np.sin(t)**2 + gxy*np.sin(t)*np.cos(t)

# unit: mm
Lx, Ly = 200, 350

dxdy = Lx/Ly
alpha = np.arctan(dxdy)
theta = alpha + np.pi/2
print(f'      alpha = {np.rad2deg(alpha):.2f} Deg')
print(f'      theta = {np.rad2deg(theta):.2f} Deg')

espaplha = extens(exxP,eyyP,exyP,theta)
print(f' esp(aplha) = {espaplha*1e6:.1f} micro-def')
```

#### b) Quais serão as extensões principais ?<a name="B"></a>

([enunciatum](#PIII67))

As componentes $\varepsilon_{xx}$ e $\varepsilon_{yy}$ são extensões principais pois no plano $S(O,x,y)$ a componente de tensão de corte é nula: $\varepsilon_{xy} = 0$.

A matriz diagonal do tensor das deformações vem:

\begin{equation*}
\underline{\underline{\varepsilon}}_\mathrm{P} = \varepsilon_\mathrm{P} =
\begin{bmatrix}
\varepsilon_\mathrm{I} & 0 & 0 \\
0 & \varepsilon_\mathrm{II} & 0 \\
0 & 0 & \varepsilon_\mathrm{III} \\
\end{bmatrix} =
\begin{bmatrix}
600.0 & 0 & 0 \\
0 & 275.0 & 0 \\
0 & 0 & -375.0 \\
\end{bmatrix}\quad\mu\varepsilon
\end{equation*}

```python
eI, eII, eIII = eyyP, exxP, ezzP

print('\nComponentes principais de deformação no plano x-y: ---------------')
print('eI = ',eI*1e6,' micro-def')
print('eII = ',eII*1e6,' micro-def')
print('eIII = ',eIII*1e6,' micro-def')
```

#### c) Qual será a distorção máxima ?<a name="C"></a>

([enunciatum](#PIII67))

A dirtorção de corte máxima pode ser obtida de acordo com as leis de transformação do tensor das deformações ou pela circunferência de Mohr. Para este último resulta uma distorção de corte máxima avaliada geometricamente como o raio do círculo de Mohr em cada plano considerado:

\begin{equation*}
\begin{aligned}
R_{xy} = \frac{1}{2}\gamma_\mathrm{max}^{xy} &= \frac{1}{2}(\varepsilon_\mathrm{I} - \varepsilon_\mathrm{II})
= \frac{1}{2}(\varepsilon_{yy} - \varepsilon_{xx}) \\
R_{xz} = \frac{1}{2}\gamma_\mathrm{max}^{xz} &= \frac{1}{2}(\varepsilon_\mathrm{I} - \varepsilon_\mathrm{III})
= \frac{1}{2}(\varepsilon_{yy} - \varepsilon_{zz}) \quad\longleftarrow\quad \gamma_\mathrm{max}\\
R_{xz} = \frac{1}{2}\gamma_\mathrm{max}^{yz} &= \frac{1}{2}(\varepsilon_\mathrm{II} - \varepsilon_\mathrm{III})
= \frac{1}{2}(\varepsilon_{xx} - \varepsilon_{zz}) \\
\end{aligned}
\end{equation*}

```python
gxymax = eI - eII
print(f'Tensão de corte máxima (x-y) = gxy\\2 = {gxymax*1e6:.1f} micro-def')
gxzmax = eI - eIII
print(f'Tensão de corte máxima (x-z) = gxz\\2 = {gxzmax*1e6:.1f} micro-def')
gyzmax = eII - eIII
print(f'Tensão de corte máxima (y-z) = gyz\\2 = {gyzmax*1e6:.1f} micro-def')
```

```python
def circleradius(D):
    e11 = D[0,0]
    e12 = D[0,1]
    e22 = D[1,1]

    center = [(e11 + e22)/2.0, 0.0]
    radius = np.sqrt((e11 - e22)**2/4.0 + e12**2)

    return center, radius

limits = np.array([[-.05e-2,.075e-2],[-.075e-2,.075e-2]])

eI, eII, eIII = eyyP, exxP, ezzP

e11 = eI; e22 = eII; e12 = 0
center1, radius1  = circleradius(np.array([[e11, e12],[e12, e22]]))

e11 = eI; e22 = eIII; e12 = 0
center2, radius2  = circleradius(np.array([[e11, e12],[e12, e22]]))

e11 = eII; e22 = eIII; e12 = 0
center3, radius3  = circleradius(np.array([[e11, e12],[e12, e22]]))

circ1 = plt.Circle((center1[0],0), radius1, facecolor='b')
circ2 = plt.Circle((center2[0],0), radius2, facecolor='r')
circ3 = plt.Circle((center3[0],0), radius3, facecolor='g')

plt.axis('image'); ax = plt.gca()
plt.gcf().gca().add_artist(circ2)
plt.gcf().gca().add_artist(circ1)
plt.gcf().gca().add_artist(circ3)
plt.xlabel(r"$\varepsilon$", size=18)
plt.ylabel(r"$\gamma/2$", size=18)
plt.text(1.05*eI, 0, r'$\varepsilon_\mathrm{I}$')
plt.text(1.2*eII, 0, r'$\varepsilon_\mathrm{II}$')
plt.text(-1.5*eII, 0, r'$\varepsilon_\mathrm{III}$')
plt.xlim(limits[0,0],limits[0,1])
plt.ylim(limits[1,0],limits[1,1])
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

#### d) Qual será a tensão tangencial ao longo da diagonal representada ?<a name="D"></a>
([enunciatum](#PIII67))

- Obtenção tensão de corte na diagonal pela lei constitutiva

A tensão de corte na diagonal representada pode ser obtida pela seguinte lei de transformação das componentes do tensor das deformações, combinada com a lei de Hooke:

\begin{equation*}
\gamma_{x1y1} = - (\varepsilon_{xx} - \varepsilon_{yy}) \sin2\theta + \gamma_{xy} \cos2\theta
\quad\wedge\quad
\tau_{x1y1} = \gamma_{x1y1} G
\end{equation*}

```python
print('--------------------------')
print('Versão 1:')
print('--------------------------')
gxyP = 2*exyP
gamx1y1 = - (exxP - eyyP)*np.sin(2*theta) + gxyP*np.cos(theta)
print('gamx1y1 = ','{:.3e}'.format(gamx1y1))
taux1y1 = G*gamx1y1
print('taux1y1 (MPa) = ','{:.3f}'.format(taux1y1))
```

- Obtenção tensão de corte na diagonal pelo círculo de Mohr

Em alternativa e pelo círculo de Mohr pode calcular-se:

\begin{equation*}
R = \frac{1}{2}(\sigma_\mathrm{I} - \sigma_\mathrm{II}) = \frac{1}{2}(\sigma_{yy} - \sigma_{xx})
\quad\wedge\quad
\tau_{x1y1} = R \sin2\alpha
\end{equation*}


Nota: Neste caso o círculo de Mohr é resolvido seguindo um caminho oposto ao habitual. Pelo conhecimento das tensões principais é fácil descrever geometricamente o círculo pelas duas abscissas ($\sigma_\mathrm{I}, \sigma_\mathrm{II}$). Desta forma, o raio do círculo e o seu centro geométrico ficam completamente definidos. É depois possível desenhar um triângulo reto para a definição do ponto no círculo correspondente ao estado de tensão no referêncial $x'y'$: o cateto oposto é a ordenada correspondente ao valor da tensão de corte $\tau_{x1y1}$; o ângulo no espaço de Mohr é duas vezes o ângulo físico: $2\alpha$.


```python
print('\n--------------------------')
print('Versão 2:')
print('--------------------------')
sI, sII = sY, sX
R = (sI - sII)/2
print(f'R = {R:.1f} (MPa)')
ang = float(2*alpha)
tauMohr = R*np.sin(ang)
print('2*alpha = ','{:.2f}'.format(np.rad2deg(ang)), '(Deg)')
print(f'|tauxy:Mohr|  = {tauMohr:.3f} (MPa)')
```

Mais concretamente, pelas leis de transformação em estado plano de tensão, o estado num elemento infinitésimal cujas faces paralela e perpendicular à diagonal do retangulo

```python
def TensorTensoes(s11,s12,s13,s22,s23,s33):
    s21 = s12; s31 = s13; s32 = s23
    return np.array([[s11,s12,s13], [s21,s22,s23], [s31,s32,s33]])

def sxx1(sxx,syy,txy,t):
    return (sxx+syy)/2 + (sxx-syy)/2*np.cos(2*t) + txy*np.sin(2*t)

def syy1(sxx,syy,txy,t):
    return (sxx+syy)/2 - (sxx-syy)/2*np.cos(2*t) - txy*np.sin(2*t)

def txy1(sxx,syy,txy,t):
    return (syy-sxx)/2*np.sin(2*t) + txy*np.cos(2*t)

def angPrincipal(sxx,syy,txy):
    return np.arctan(2*txy/(sxx-syy))/2

def angS(s11,s22,s12): return -np.arctan((s11-s22)/(2*s12))

print('-----------------')
print('PLANO x-y:')
print('-----------------')
tt0 = mtP
sxx = tt0[0,0]
syy = tt0[1,1]
txy = tt0[0,1]

angt = np.arange(-45,145,1)
tsxx = sxx1(sxx,syy,txy,np.deg2rad(angt))
tsyy = syy1(sxx,syy,txy,np.deg2rad(angt))
ttxy = txy1(sxx,syy,txy,np.deg2rad(angt))

print('Tensões (normais) principais: -------------------------------------------')
aprinc = np.rad2deg(angPrincipal(sxx,syy,txy))
print('angPrincipal (deg) =','{:.2f}'.format(aprinc))
tsxx1 = sxx1(sxx,syy,txy,np.deg2rad(aprinc))
tsyy1 = syy1(sxx,syy,txy,np.deg2rad(aprinc))
ttxy1 = txy1(sxx,syy,txy,np.deg2rad(aprinc))
print('tens xx - plano principal [1] =','{:.2f}'.format(tsxx1),' MPa')
print('tens yy - plano principal [1] =','{:.2f}'.format(tsyy1),' MPa')
print('tens xy - plano principal [1] =','{:.2f}'.format(ttxy1),' MPa')

aprinc2 = aprinc + 90
print('\nangPrincipal 2 (deg) =','{:.2f}'.format(aprinc2))
tsxx2 = sxx1(sxx,syy,txy,np.deg2rad(aprinc2))
tsyy2 = syy1(sxx,syy,txy,np.deg2rad(aprinc2))
ttxy2 = txy1(sxx,syy,txy,np.deg2rad(aprinc2))
print('tens xx - plano principal [2] =','{:.2f}'.format(tsxx2),' MPa')
print('tens yy - plano principal [2] =','{:.2f}'.format(tsyy2),' MPa')
print('tens xy - plano principal [2] =','{:.2f}'.format(np.abs(ttxy2)),' MPa')

print('\nTensões de corte maximas/mínima: -------------------------------------------')
aS1 = angS(sxx,syy,txy)/2 # ângulo no espaço físico, radianos
print('Ângulo tensão corte max (Deg) =','{:.2f}'.format(np.rad2deg(aS1)))
tcortexxMa = sxx1(sxx,syy,txy,aS1)
tcorteyyMa = syy1(sxx,syy,txy,aS1)
tcortexyMa = txy1(sxx,syy,txy,aS1)
print('tens xx - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcortexxMa))
print('tens yy - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcorteyyMa))
print('tens xy - plano  tensão de corte max (MPa) =','{:.2f}'.format(tcortexyMa))

aS2 = aS1 + np.pi/2
print('Ângulo tensão corte min (Deg) =','{:.2f}'.format(np.rad2deg(aS2)))
tcortexxMb = sxx1(sxx,syy,txy,aS2)
tcorteyyMb = syy1(sxx,syy,txy,aS2)
tcortexyMb = txy1(sxx,syy,txy,aS2)
print('tens xx - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcortexxMb))
print('tens yy - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcorteyyMb))
print('tens xy - plano  tensão de corte min (MPa) =','{:.2f}'.format(tcortexyMb))

# use LaTeX fonts in the plot
plt.rc('text', usetex=False)
plt.rc('font', family='serif')

fig, ax = plt.subplots(figsize=(7,5))
ax.plot(angt, tsxx, 'k--', label=r'$\sigma_{xx}$', linewidth=3)
ax.plot(angt, tsyy, 'k:', label=r'$\sigma_{yy}$', linewidth=3)
ax.plot(angt, ttxy, 'k', label=r'$\tau_{xy}$', linewidth=3)
# points: principal stresses
ax.plot(aprinc, tsxx1, 's'), ax.text(aprinc, tsxx1,r'$\sigma_{xx}$')
ax.plot(aprinc, tsyy1, 's'), ax.text(aprinc, tsyy1,r'$\sigma_{yy}$')
ax.plot(aprinc, ttxy1, 's'), ax.text(aprinc, ttxy1,r'$\sigma_{xy}$')
ax.plot(aprinc2, tsxx2, 's'), ax.text(aprinc2, tsxx2,r'$\sigma_{xx}$')
ax.plot(aprinc2, tsyy2, 's'), ax.text(aprinc2, tsyy2,r'$\sigma_{yy}$')
ax.plot(aprinc2, ttxy2, 's'), ax.text(aprinc2, ttxy2,r'$\tau_{xy}$')
# points: critical shear stresses
ax.plot(np.rad2deg(aS1), tcortexxMa, 'v')
ax.text(np.rad2deg(aS1), tcortexxMa,r'$\varepsilon_{xx}$')
ax.plot(np.rad2deg(aS1), tcorteyyMa, 'v')
ax.text(np.rad2deg(aS1), tcorteyyMa,r'$\varepsilon_{yy}$')
ax.plot(np.rad2deg(aS1), tcortexyMa, 'v')
ax.text(np.rad2deg(aS1), tcortexyMa,r'$\gamma_{xy}$')
ax.plot(np.rad2deg(aS2), tcortexxMb, 'v')
ax.text(np.rad2deg(aS2), tcortexxMb,r'$\varepsilon_{xx}$')
ax.plot(np.rad2deg(aS2), tcorteyyMb, 'v')
ax.text(np.rad2deg(aS2), tcorteyyMb,r'$\varepsilon_{yy}$')
ax.plot(np.rad2deg(aS2), tcortexyMb, 'v')
ax.text(np.rad2deg(aS2), tcortexyMb,r'$\gamma_{xy}$')

ax.set(xlabel='Ângulo (deg)', ylabel='Transformação das tensões')
legend = ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., shadow=True, fontsize=18)
ax.axhline(y=0, color='k', linewidth=1)
ax.axvline(x=0, color='k', linewidth=1)
ax.grid()
plt.show()
```

```python
def mohr(S):
    S11 = S[0][0]
    S12 = S[0][1]
    S22 = S[1][1]
    center = [(S11 + S22)/2.0, 0.0]
    radius = np.sqrt((S11 - S22)**2/4.0 + S12**2)
    Smin = center[0] - radius
    Smax = center[0] + radius

    print("  Tensão normal mínima: ", '{:.1f}'.format(Smin))
    print("  Tensão normal máxima: ", '{:.1f}'.format(Smax))
    print("   Tensão normal média: ", '{:.1f}'.format(center[0]))
    print("Tensão de corte mínima: ", '{:.1f}'.format(-radius))
    print("Tensão de corte máxima: ", '{:.1f}'.format(radius))

    print('Ponto A (','{:.1f}'.format(S11),',','{:.1f}'.format(-S12),')')
    print('Ponto B (','{:.1f}'.format(S22),',','{:.1f}'.format(S12),')')

    circ = plt.Circle((center[0],0), radius, facecolor='tab:olive', lw=3,
    edgecolor='tab:orange')
    desv = 0.1*radius
    plt.axis('image')
    ax = plt.gca()
    ax.add_artist(circ)
    ax.set_xlim(Smin - .3*radius, Smax + .3*radius)
    ax.set_ylim(-1.3*radius, 1.3*radius)
    plt.plot([S11, S22], [-S12, S12], 'ko')
    plt.plot([S11, S22], [-S12, S12], 'k')
    plt.plot(Smin, 0, 'bs')
    plt.plot(Smax, 0, 'bs')
    plt.plot(center[0], radius, 'rv')
    plt.plot(center[0], -radius, 'rv')
    plt.plot(center[0], center[1], 'o', mfc='w')
    plt.text(center[0]+desv, desv, 'C')
    plt.text(S11 + desv, -S12, 'A')
    plt.text(S22 + desv, S12, 'B')
    plt.text(Smin - 0.25*radius, -0.15*radius, r'$\sigma_\mathrm{min}$')
    plt.text(Smax + desv, -0.15*radius, r'$\sigma_\mathrm{max}$')
    plt.text(center[0]-4*desv, 1.1*radius, r'$\tau_\mathrm{max}$')
    plt.text(center[0]-4*desv, -1.1*radius, r'$\tau_\mathrm{min}$')
    plt.xlabel(r"$\sigma$", size=18)
    plt.ylabel(r"$\tau$", size=18)
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    plt.show()

    return Smin, Smax, center[0],radius,-radius

print('Círculo de Mohr - PLANO x-y:\n')
print('unidades: MPa')
sxx = (sI + sII)/2 + (sI - sII)/2*np.cos(2*alpha)
print('Sx1x1 = ','{:.3f}'.format(sxx))
syy = (sI + sII)/2 - (sI - sII)/2*np.cos(2*alpha)
print('Sy1y1 = ','{:.3f}'.format(syy))
sxy = taux1y1
print('tx1y1 = ','{:.3f}'.format(sxy), '\n')
Tmin, Tmax, Tmed, Cmax, Cmin = mohr(np.array([[sxx, sxy],[sxy, syy]]))
```
