Os dois veios maciços de aço ($G = 77$ GPa) representados estão ligados por engrenagens. Os dois veios são suportados por três chumaceiras como se indica e a extremidade D está fixa. Sabendo que o raio da engrenagem B é $r_B$ = 20 mm, determine:

[a)](#A) O valor da tensão tangencial máxima em cada veio e a rotação da secção A quando um momento de torção $T_A = 80$ N.m é aplicado nessa extremidade.

[b)](#B) Qual deverá ser o raio da engrenagem B, $r_B$, para que a tensão tangencial máxima nos dois veios seja igual.

### Resolução

#### (a) O valor da tensão tangencial máxima em cada veio e a rotação da secção A quando um momento de torção $T_A = 80$ N.m é aplicado nessa extremidade.

```python
import numpy as np
import matplotlib.pyplot as plt

# units: Pa, m, N
rC, rB = 60.e-3, 20.e-3
G = 77e9
dCD, dAB = 24.e-3, 20.e-3
rCD, rAB = dCD/2, dAB/2
LCD, LAB = 0.4, 0.5
TA = 80.

def J0(r):
    return np.pi*r**4/2

def rot(T,L,J,G):
    return T*L/J/G
```

**Equilíbrio estático (momentos)**

\begin{equation*}
\sum M_B = 0 \quad\Rightarrow\quad T_A = T_B = F r_B
\end{equation*}

```python
F = TA/rB
print(f'{F/1e3:.2f} kN')
```


\begin{equation*}
\sum M_C = 0 \quad\Rightarrow\quad T_C = T_D = F r_C
\end{equation*}

```python
TC = F*rC
print(f'{TC:.2f} N')
```

**Barra AB**

\begin{equation*}
\tau_\textrm{max,AB} = \frac{T_\textrm{A} C_\textrm{AB}}{J_\textrm{O,AB}}
\end{equation*}

```python

J0AB = J0(rAB)

tauMaxAB = TA*rAB/J0AB
print(f'{tauMaxAB/1e6:.2f} MPa')
```

**Barra CD**

\begin{equation*}
\tau_\textrm{max,CD} = \frac{T_\textrm{D} C_\textrm{CD}}{J_\textrm{O,CD}}
\end{equation*}

```python

J0CD = J0(rCD)
TD = TC
tauMaxCD = TD*rCD/J0CD
print(f'{tauMaxCD/1e6:.2f} MPa')
```

**Ângulo de rotação**

\begin{equation*}
\phi_\textrm{C} = \frac{T_\textrm{D}L_\textrm{CD}}{J_\textrm{O,CD}G}
\end{equation*}


```python
phiC = rot(TD,LCD,J0CD,G)
print(f'{phiC:.4f} rad')
```

**Cinemática da Engrenagem**

\begin{equation*}
V_\textrm{C} = V_\textrm{C}
\quad\wedge\quad
w_\textrm{C} r_\textrm{C} = w_\textrm{B} r_\textrm{B}
\quad\wedge\quad
\phi_\textrm{C} r_\textrm{C} = \phi_\textrm{B} r_\textrm{B}
\end{equation*}


\begin{equation*}
\phi_\textrm{B} = \frac{r_\textrm{C}}{r_\textrm{B}}\phi_\textrm{C}
\end{equation*}

```python
phiB = rC/rB*phiC
print(f'{phiB:.4f} rad')
```

**Rotaçao $\phi_A$**

\begin{equation*}
\phi_\textrm{A/B} = \phi_A - \phi_B =  \frac{T_\textrm{A} L_\textrm{AB}}{J_\textrm{O,AB} G}
\end{equation*}




```python
phiAB = rot(TA,LAB,J0AB,G)
print(f'{phiAB:.4f} rad')
```

**Rotaçao $\phi_A$**

\begin{equation*}
\phi_A = \phi_\textrm{A/B} + \phi_B
\end{equation*}

```python
phiA = phiAB + phiB
print(f'{phiA:.4f} rad = {np.rad2deg(phiA):.3f}º')
```

### (b) Qual deverá ser o raio da engrenagem B, $r_B$, para que a tensão tangencial máxima nos dois veios seja igual.

\begin{equation*}
\tau_\textrm{max,AB} = \tau_\textrm{max,CD}
\quad\leftrightarrow\quad
\frac{T_\textrm{A} c_\textrm{AB}}{J_\textrm{O,AB}} = \frac{T_\textrm{D} c_\textrm{CD}}{J_\textrm{O,CD}}
\end{equation*}

\begin{equation*}
T_\textrm{A} =
\left(\frac{J_\textrm{O,AB}}{J_\textrm{O,CD}}\right)
\left(\frac{c_\textrm{CD}}{c_\textrm{AB}}\right)
T_\textrm{D} =
\left(\frac{r_\textrm{AB}}{r_\textrm{CD}}\right)³
T_\textrm{D}
\end{equation*}

```python
TDnew = TA/(rAB/rCD)**3
print(f'{TDnew:.2f} N.m')
```

**Equilíbrio de momentos**

\begin{equation*}
T_\textrm{D} = r_\textrm{C} F
\end{equation*}

```python
Fnew = TDnew/rC
print(f'Fnew = {Fnew:.2f} m')
```

\begin{equation*}
T_\textrm{A} = r_\textrm{A} F
\end{equation*}

```python
rAnew = TA/Fnew
print(f'rAnew = {rAnew:.3f} m = {rAnew*1e3:.3f} mm')
```
