## Problema 2 <a name="PIV82"></a>

As Figuras representam um tubo de aço, $E= 210$ GPa, $\alpha = 11,7 \times 10^{-6} / ^\circ\!C $, construído a partir de uma chapa enrolada e soldada. A soldadura, em forma de hélice, faz 33$^\circ$ com o plano perpendicular ao eixo do tubo, e suporta uma tensão tangencial máxima $\tau = 35$ MPa e uma tensão normal $\sigma = 70$ MPa. Sabe-se que o tubo tem um comprimento $L = 200$ mm, um diâmetro exterior de 150 mm e uma espessura de 6 mm. Considerando que, à temperatura de 20$^\circ\!C$, o tubo está perfeitamente ajustado entre as duas superfícies $A$ e $B$, determine a temperatura máxima a que o tubo poderá ser aquecido sem danificar a soldadura.

## Resolução

alongamento de uma barra provocado pela variação de temperatura:

\begin{equation*}
\delta_T = \alpha (\Delta T) L
\end{equation*}

assoicado ao alongamento ocorrerá uma deformação


\begin{equation*}
\varepsilon = \frac{\delta_T}{L} = \alpha (\Delta T)
\end{equation*}


Problema estaticamente indeterminado:

- i) Estrutura isostática equivalente : tudo encastrado apenas num dos seus apoios e com a outra extremidade livre

- ii) tubo encastrado numa das extremidades e livre na outra, sofre um alongamento introduzido pela variação de
temperatura: $+\delta_T$ (proporcional a uma tensão térmica de tração - expansão)

- iii) Configuração isostática equivalente submetida à ação de uma força externa $P$, designada por redundante, e que
representa a reação nesse encastramento do cilindro, $-\delta_P$ (proporcional a uma tensão de compressão)

- iv) Equação de compatibilidade de deslocamentos:

\begin{equation*}
\begin{aligned}
\delta_T - \delta_P &= 0 \\
\alpha \Delta T L - \frac{PL}{AE} &= 0 \\
P &= AE \alpha \Delta T
\end{aligned}
\end{equation*}

O estado uniaxial pode ser escrito no estao de tensão $(\sigma,\tau)$ num plano oblíquo sob o carregamento axial:

\begin{equation*}
\begin{aligned}
\sigma &= \frac{P}{A}\cos^2\theta\\
\tau &= \frac{P}{A}\sin\theta\cos\theta
\end{aligned}
\end{equation*}

- $T^\mathrm{max} \leftarrow \sigma^\mathrm{max}$:

\begin{equation*}
\Delta T = T- T_0  = \frac{\sigma^\mathrm{max}}{E \alpha \cos^2\theta}
\end{equation*}


- $T^\mathrm{max} \leftarrow \tau^\mathrm{max}$:


\begin{equation*}
\Delta T = T- T_0  = \frac{\tau^\mathrm{max}}{E \alpha \sin\theta\cos\theta}
\end{equation*}
