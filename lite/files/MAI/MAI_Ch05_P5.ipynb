### Resolução

#### a) Determine a coordenada $x$ do centroide da superfície indicada em função de $r$.

comece-se por decompor a figura original em duas figuras mais simples conforme  se mostra pela figura abaixo.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P5/MAI_Ch05_P5a.jpg"
width="900">

Esta decomposição permite determinar o centroide da área total como a combinação linear dos centroides das sub-áreas usadas na sua definição:

\begin{equation*}
\overline{x} = \sum_{i} \frac{A_i }{A} \overline{x}_i
= \frac{A_1}{A} \overline{x}_1 - \frac{A_2}{A} \overline{x}_2
\end{equation*}

A área e os centroides das sub-áreas elementares são obtidos diretamente de formulário por consulta de tabelas próprias:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P5/MAI_Ch05_P5b.jpg"
width="900">

**Triângulo**

Das tabelas, para uma área triangular genérica têm-se diretamente para a sub-área (1):

| Área, $A_1$ | Centroide, $\overline{x}_1$ |
|:----------------:|:------:|
|$\displaystyle\frac{r^2}{2}$|$\displaystyle\frac{r}{3}$|

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

r = sy.symbols('r')

A1 = r**2/2
print(f'A1 = {A1} m2')

x1 = r/3
print(f'x1 = {x1} m')
```

**Setor circular**

Das tabelas (setor circular), para a sub-área (2), que representa metade do quarto de círculo, têm-se:

_Área_

\begin{equation*}
\begin{aligned}
A_2 &= \alpha r^2
~\wedge~ \alpha =
\frac{\theta}{2}
~\wedge~ \theta = 45^\circ
~\wedge~ \alpha =
\frac{45^\circ}{2} = 22.5^\circ = \frac{\pi/4}{2} =
\frac{\pi}{8} = 0.393~
\textrm{rad}
\end{aligned}
\end{equation*}

\begin{equation*}
\begin{aligned}
~\therefore~
A_2 = \frac{\pi r^2}{8} \\
\end{aligned}
\end{equation*}

_Centroide_

Pelo esquema do problema escreve-se:

\begin{equation*}
\begin{aligned}
\overline{x}_2 &= \overline{OC}\cos\left(\frac{\pi}{4}+\alpha\right)
= \overline{OC}\cos\left(\frac{\pi}{2}-\alpha\right)
= \overline{OC}\sin\left(\alpha\right)  \\
\end{aligned}
\end{equation*}

Por consulta direta da tabela para a superfície em análise,

\begin{equation*}
\overline{OC} = \frac{2r\sin\alpha}{3\alpha} \\
\end{equation*}

Pelo que,

\begin{equation*}
\begin{aligned}
x_2 &= \frac{2r\sin^2\alpha}{3\alpha}
= \frac{16r}{3\pi}\sin^2\left(\frac{\pi}{8}\right)
= \frac{16r}{3\pi}\left[\frac{1}{2}\sqrt{2-\sqrt{2}}\right]^2
\end{aligned}
\end{equation*}

- formulário:

    \begin{equation*}
    \sin\left(\frac{\pi}{8}\right) = \frac{1}{2}\sqrt{2-\sqrt{2}}
    \end{equation*}

```python
alpha = np.deg2rad(45)/2 # unit: radianos
print(f'alpha (= pi/8) = {alpha:.3f} radianos')
print(f'sin(pi/8)      = {np.sin(alpha):.3f} radianos')

A2 = alpha*r**2
print('A2  = ',A2)

OC2 = 2*r*np.sin(alpha)/3/alpha
print('OC2 = ',OC2)

x2 = OC2*np.sin(alpha)
print(f'Calc. 1: x2 = {x2}')

x2b = OC2*np.cos(np.pi/4+alpha)
print(f'calc. 2: x2 {x2b}')

x2c = (16*r/3/np.pi)*(0.5*np.sqrt(2-np.sqrt(2)))**2
print(f'calc. 3: x2 = {x2c}')
```

|  Sub-área | `(1)` | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;`(2)` &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;| $\sum: (1) - (2)$ |
|:----------:|:----------:|:----------:|:----------:|
| $A_i$, m $^2$ | $\displaystyle\frac{r^2}{2}$ | $\displaystyle\frac{\pi r^2}{8}$ | $0.107r^2$ |
| $\overline{x}_i$, m | $\displaystyle\frac{r}{3}$ | $\displaystyle\frac{16r}{3\pi}\left[\frac{1}{2}\sqrt{2-\sqrt{2}}\right]^2$ |
| $\overline{x}_iA_i$, m $^3$ | $\displaystyle\frac{r^3}{6}$ | $\displaystyle\frac{2r^3}{3}\left[\frac{1}{2}\sqrt{2-\sqrt{2}}\right]^2$ | $0.069r^3$ |

com,

\begin{equation*}
\overline{x} = \frac{\sum_i \overline{x}_i A_i}{\sum_i A_i}
= \frac{0.069r^3}{0.107r^2} = 0.643 r \,\textrm{(m)}
\end{equation*}

```python
A = A1 - A2
print('A    = ',A)

SxiAi = x1*A1 - x2*A2
print('SxiAi = ',SxiAi)

xcent = SxiAi/A
print('xcent = ',xcent)
```

#### b)  Enuncie o 2º Teorema de Pappus-Guldinus.

Os dois teoremas conhecidos por Pappus-Guldinus, foram inicialmente formulados por Pappus na grécia durante o século III d.C. e posteriormente reformulado pelo matemático suíço Guldinus (1577-1643)

**Teorema II**

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/GFormulas/appus-GuldinusiI.jpg"
width="500">

O volume de um corpo de revolução é igual ao produto da área da superfície geratriz pelo caminho percorrido pelo centroide dessa superfície durante o movimento de rotação que gera o corpo. Matematicamente, este teorema conduz a:

\begin{equation*}
V = 2\pi\,\overline{x}\,A
\end{equation*}

em que $2\pi\,\overline{x}$ representa, por outras palavras, o perímetro de uma circunferência de raio $\overline{x}$.



#### c) Sabendo que o volume do sólido de revolução gerado pela rotação da superfície
a tracejado em torno do eixo $y$ é de 750 cm $^3$, determine o valor de $r$.

Recorrendo ao 2º Teorema de Pappus-Guldinus, pode-se determinar o valor do raio $r$ a partir do volume do sólido.

\begin{equation*}
V = 2\pi\,\overline{x}\,A = 2\pi (0.643 r)(0.107r^2)
= 0.434 r^3
\end{equation*}

```python
V = 2*np.pi*xcent*A
print('V = ',V)
```

mas como $V = 750 $ cm $^3$, resulta finalmente:

\begin{equation*}
r = (V/0.434)^{1/3}
\end{equation*}

```python
Vn = 750*1e-6 # unit: m3

eq = V - Vn
sol = solve(eq,r)
print('raiz 1: ',sol[0])
print('raiz 2 (n. complexo): ',sol[1])
print('raiz 2 (n. complexo): ',sol[2],'\n')
rn = sol[0]
print('r = ',rn,' m')
```

#### Anexo: formulário

A primeira etapa na caracterização das propriedades de área passa pela localização do centroide da secção. O centroide é o ponto geométrico sobre a qual a área da secção está igualmente distribuída. Por definição, o primeiro momento de área sobre os eixos $x$ e $y$, é dado, respetivamente, por:

\begin{equation*}
Q_x = \int_A y\,\mathrm{d}A~,
\quad Q_y = \int_A x\,\mathrm{d}A
\end{equation*}

O centroide de uma área A é o ponto C definido pelas coordenas $\overline{x}$ e $\overline{y}$, que satisfaz as seguintes condições.

\begin{equation*}
\overline{x} = \frac{Q_y}{A} = \frac{\displaystyle\int_A x\,\mathrm{d}A}{\displaystyle\int_A\mathrm{d}A}~,
\quad
\overline{y} = \frac{Q_x}{A} = \frac{\displaystyle\int_A y\,\mathrm{d}A}{\displaystyle\int_A\mathrm{d}A}
\end{equation*}

Quando uma secção contem um eixo de simetria, o centroide está contido nesse eixo, na medida em que o primeiro momento de área em relação a um eixo de simetria é zero. Quando uma secção contem dois planos de simetria, o centroide é definido pelo ponto de interseção destes dois eixos.

Frequentemente, a área de uma secção pode ser decomposta em formas geométricas simples (e.g., retangulares, círculos ou triângulos), para as quais a área e as coordenadas do centroide são mais fáceis de determinar. Quando uma área pode ser reconstruida de $n$ elementos, a área total pode ser obtida pela soma algébrica das sub-áreas individuais, e o momento resultante em relação a um eixo pode ser obtido pela soma algébrica dos momentos das áreas individuais. Assim, o centroide de uma área composta pode ser determinado por:

\begin{equation*}
\overline{x} = \frac{\displaystyle\sum A_i \overline{x}_i}{\displaystyle\sum
A_i},\quad
\overline{y} = \frac{\displaystyle\sum A_i \overline{y}_i}{\displaystyle\sum A_i}
\end{equation*}

em que $\overline{x}_i$ e $\overline{y}_i$ representam as coordenadas do centroide de cada componente de área $A_i~(i=1,\cdots,n)$.
