#### Problema 11

Considere na figura o referencial e a manivela onde está montado um sarilho. O mecanismo tem
dois pontos de apoio, A (rótula esférica) e B (chumaceira de cargas radiais), e no sarilho está
suspenso um peso de 150 N. Despreze o peso do mecanismo e o atrito. Calcule a intensidade da
força horizontal $P$ para manter a manivela estática na posição representada e as
correspondentes componentes de reação nos apoios. Justifique todos os cálculos e inclua o
respetivo diagrama de corpo livre.

#### Resolução




\begin{equation}
M_{AB} = \vec{\lambda}_{AB} \cdot \vec{M}_B^F = 10~\textrm{N.m}
\end{equation}



\begin{equation}
M_{AB} = \vec{\lambda}_{AB} \cdot \vec{M}_B^R
= \vec{\lambda}_{AB} \cdot (\vec{M}_B^F + \vec{M}_B^T +
\vec{M}_B^{R_A})
\end{equation}

\begin{equation}
= \underbrace{\vec{\lambda}_{AB} \cdot (\vec{r}_{AB} \times \vec{R}_{A})}_{0}
+ \underbrace{\vec{\lambda}_{AB} \cdot \vec{M}_B^F}_{10}
+ \vec{\lambda}_{AB} \cdot \vec{M}_B^T = 0
\end{equation}


\begin{equation}
\vec{\lambda}_{AB} \cdot \vec{M}_B^T =
\vec{\lambda}_{AB} \cdot (\vec{r}_{CB} \times \vec{T}_{CD})
\end{equation}

\begin{equation}
\vec{r}_{CD} = r_x \vec{\imath} + r_y \vec{\jmath}
\end{equation}

\begin{equation}
\vec{T}_{CD} = T_x \vec{\imath} + T_y \vec{\jmath} - T_z \vec{k}
\end{equation}
