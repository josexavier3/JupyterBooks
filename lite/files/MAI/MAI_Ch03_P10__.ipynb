#### Problema 10

A figura representa uma barra AB homogénea de massa $m= 3$ kg que se encontra articulada em B e apoiada em A na superfície do bloco de massa $m_0 = 3$ kg, de tal modo que $\theta = 30^\circ$. O bloco por sua vez encontra-se sobre uma superfície horizontal. Sabendo que há atrito entre as superfícies de contacto (barra – bloco e bloco – superfície horizontal), $\mu_e$ = 0,6; $\mu_c$ = 0,5

**b)** Para o caso em que $P =$ 100 N, determine o valor da força de atrito entre as superfícies.
