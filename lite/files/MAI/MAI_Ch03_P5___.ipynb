#### Problema 5

A figura representa duas massas $m_1$ e $m_2$ assentes num plano inclinado e ligadas por um cabo passando na roldana A. Sabe-se que a massa $m_1 = 40$ kg e $m_2 = 60$ kg e o coeficiente de atrito estático entre todas as superfícies em contacto é $\mu_\mathrm{e}$ = 0,5. Responda às seguintes alíneas:

**a)** Trace separadamente os diagramas de corpo livre de cada uma das massas e da roldana.

**b)** Determine a força mínima $\vec{P}$ que deverá ser aplicada para iniciar o movimento descendente do bloco de massa $m_2$. Determine ainda a correspondente força de tração no cabo.
