Identificam-se as incógnitas do problema:

1. No apoio A: $R_{Ax}$, $R_{Ay}$ e  $R_{Az}$ (3 reações de apoio);
2. No apoio B: $R_{By}$ e $R_{Bz}$ (2 reações de apoio);
3. Força $P$ aplicada no Ponto D (força atuante).

O equilíbrio do corpo rígido no espaço conduz a 6 equações, pelo que, existindo 6 incógnitas, o problema conduz a um sistema de equações bem determinado e com solução única.

#### b) Calcule o momento resultante $M_0$ das forças P e T em relação à origem do referencial situada a  meio do veio.

O momento resultante $M_0$ das forças $P$ e $T$ em relação à origem do referencial (situado a meio do veio) é dado por:

\begin{equation*}
\vec{M}_O = \sum_{i}(\vec{r}_\textrm{Oi}\times\vec{F}_\textrm{i})
= (\vec{r}_{\textrm{OC}}\times\vec{T}) +
(\vec{r}_{\textrm{OD}}\times\vec{P})
\end{equation*}

com,

\begin{equation*}
\begin{aligned}
\vec{r}_{\textrm{OC}} &= x_\textrm{OC}\,\vec{\imath} +
\textrm{OC'}\sin\alpha\,\vec{\jmath} -
\textrm{OC'}\cos\alpha\,\vec{k} \\
& = 60\,\vec{\imath} +
120\sin(30\frac{\pi}{180})\,\vec{\jmath} -
120\cos(30\frac{\pi}{180})\,\vec{k} \\
& = 60\,\vec{\imath} +
60\,\vec{\jmath} -
103.9\,\vec{k}\\
\vec{r}_{\textrm{OD}} &= 0\,\vec{\imath} + 0\,\vec{\jmath} + z_\textrm{OD}\,
\vec{k} = 200\,\vec{k}
\end{aligned}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
from sympy.physics.mechanics import ReferenceFrame, Vector, dot, cross

OD = 200 # unit: mm
OCp = 120
Cx = 60
alpha = 30 # unit: graus
T = 400 # unit: N

S = ReferenceFrame('S')
xOC = Cx
yOC = OCp*np.sin(np.deg2rad(alpha))
zOC = OCp*np.cos(np.deg2rad(alpha))
rOC = xOC*S.x + yOC*S.y - zOC*S.z
print('rOC : ',rOC)
rOD = 0*S.x + 0*S.y + OD*S.z
print('rOD : ',rOD)
```

Os produtos vetoriais podem ser determinados pelo cálculo dos seguintes determinanted:

\begin{equation*}
\bullet\quad\vec{r}_{\textrm{OC}}\times\vec{T} =
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
x_\textrm{OC} & y_\textrm{OC} & z_{\textrm{OC}} \\
0 & -T & 0
\end{vmatrix} = (-41569.2)\vec{\imath} + 0\vec{\jmath} + (- 24000)\vec{k}
\end{equation*}

```python
vT = -T*S.y
CP1 = cross(rOC,vT); print('Cross 1: ',CP1)
```

\begin{equation*}
\bullet\quad\vec{r}_{\textrm{OD}}\times\vec{P} =
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
0 & 0 & z_{\textrm{OD}} \\
0 & -P & 0
\end{vmatrix} =
\vec{\imath}
\begin{vmatrix}
0 & z_{\textrm{OD}} \\
-P & 0
\end{vmatrix} =
200 P~\vec{\imath}
\end{equation*}

```python
p = sy.symbols('p')
vP = -p*S.y
CP2 = cross(rOD,vP); print('Cross 2: ',CP2)
```

Pelo que momento $\vec{M}_0$ resulta:

\begin{equation*}
\begin{aligned}
\vec{M}_O &= (\vec{r}_{\textrm{OC}}\times\vec{T}) +
(\vec{r}_{\textrm{OD}}\times\vec{P}) \\
& = (- 41569.2 + 200\,P)\vec{\imath} + 0\vec{\jmath} +
(- 24000)\vec{k}
\end{aligned}
\end{equation*}

```python
expM0 = CP1 + CP2
print('Exp M0 = ',expM0)
```

#### c) Utilize a equação de equilíbrio de momentos em relação a um eixo ($\vec{\lambda}_x \cdot \vec{M}_0 = 0$) para calcular o valor de P.

Relembrando o formulário para o momento segundo um eixo:

\begin{equation*}
\begin{aligned}
M_\textrm{veio} &=
\underbrace{\vec{\lambda}\qquad}_{\substack{\text{versor com a} \\ \text{direção do
 eixo}}}\cdot
\underbrace{\left(\sum \vec{r}\times\vec{F}\right)\qquad}_{\substack{\text{momento
resultante das forças} \\ \text{em relação a um ponto desse eixo}}}\\
&= \vec{\lambda}\cdot
(\vec{r}_{\textrm{OC}}\times\vec{T}) +
(\vec{r}_{\textrm{OD}}\times\vec{P}) \\
& = \vec{\lambda}\cdot[(- 41569.2 + 200\,P)\vec{\imath} + 0\vec{\jmath} +
(- 24000)\vec{k}]
\end{aligned}
\end{equation*}

Neste caso considere-se o eixo do veio AB ($\lambda_x$):

\begin{equation*}
\vec{\lambda} = \vec{\lambda}_\textrm{eixo} = \vec{\lambda}_x
= \frac{\vec{OB}-\vec{OA}}{||\vec{AB}||}
= \frac{200\,\vec{\imath}}{\sqrt{200^2}} = \vec{\imath}
\end{equation*}

```python
A = [-100,0,0]
B = [100,0,0]

vAB = np.subtract(B,A); print('vAB = ',vAB)

modAB = np.sqrt(np.dot(vAB,vAB))
print('modulo de AB = ','{:.1f}'.format(modAB),' mm')
```

Assim sendo, para que exista equilíbrio estático, resulta:

\begin{equation*}
\begin{aligned}
M_\textrm{veio}
&= \vec{\imath}\cdot[(- 41569.2 + 200\,P)\vec{\imath} + 0\vec{\jmath} +
(- 24000)\vec{k}] = 0\\
&~\Leftrightarrow~ - 41569.2 + 200\,P = 0\\
&~\Leftrightarrow~ P = 207.85~\textrm{N}\\
\end{aligned}
\end{equation*}

```python
lambdx = vAB[0]/modAB*S.x + vAB[1]/modAB*S.y + vAB[2]/modAB*S.z
Meixo = dot(lambdx,expM0)
print('dot() = ',Meixo)
sol = solve({Meixo},{p})
P = float(sol[p]); print('P = ','{:.2f}'.format(P),' N')
```

#### d) Calcule as reações nos apoios A e B.

**Equilíbrio dos momentos**

\begin{equation*}
\begin{aligned}
\bullet\quad\sum \vec{M}_O &= \sum (\vec{r}\times\vec{F}) \\
& = (\vec{r}_{\textrm{OC}}\times\vec{T}) +
(\vec{r}_{\textrm{OD}}\times\vec{P}) + + (\vec{r}_{\textrm{OA}}\times\vec{R}_\textrm{A}) 
+ (\vec{r}_{\textrm{OB}}\times\vec{R}_\textrm{B}) \\
&= \vec{M}_O(\vec{T}, \vec{P}) + (\vec{r}_{\textrm{OA}}\times\vec{R}_\textrm{A})
+ (\vec{r}_{\textrm{OB}}\times\vec{R}_\textrm{B}) = \vec{0}
\end{aligned}
\end{equation*}

\begin{equation*}
\begin{aligned}
\vec{r}_{\textrm{OA}}\times\vec{R}_\textrm{A} =
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} & \vert & \vec{\imath}\\
-x_{\textrm{OA}} & 0 & 0 & \vert & -x_{\textrm{OA}}  \\
R_\textrm{Ax} & R_\textrm{Ay} & R_\textrm{Az} & \vert & R_\textrm{Ax}
\end{vmatrix} =
\begin{vmatrix}
0 & -x_{\textrm{OA}}  \\
R_\textrm{Az} & R_\textrm{Ax}
\end{vmatrix}\vec{\jmath}  +
\begin{vmatrix}
-x_{\textrm{OA}} & 0 \\
R_\textrm{Ax} & R_\textrm{Ay}
\end{vmatrix} \vec{k}
\end{aligned}
\end{equation*}

\begin{equation*}
\begin{aligned}
\qquad\qquad~ = [0-(-x_{\textrm{OA}})R_\textrm{Az}] \vec{\jmath} +
[(-x_{\textrm{OA}})R_\textrm{Ay}-0]
\vec{k}
= 100R_\textrm{Az}\vec{\jmath} - 100R_\textrm{Ay}\vec{z}
\end{aligned}
\end{equation*}

```python
rax, ray, raz = sy.symbols('rax ray raz')
rOA = A[0]*S.x
vRA = rax*S.x + ray*S.y +  raz*S.z
CP3 = cross(rOA,vRA); print('Cross 3: ',CP3)
```

\begin{equation*}
\vec{r}_{\textrm{OB}}\times\vec{R}_\textrm{B}=
\begin{vmatrix}
\vec{\imath} & \vec{\jmath} & \vec{k} \\
-x_{\textrm{OA}} & 0 & 0 \\
0 & R_\textrm{By} & R_\textrm{Bz}
\end{vmatrix} = -100R_\textrm{Bz}\vec{\jmath} + 100R_\textrm{By}\vec{z}
\end{equation*}

```python
rby, rbz = sy.symbols('rby rbz')
rOB = B[0]*S.x
vRB = rby*S.y +  rbz*S.z
CP4 = cross(rOB,vRB); print('Cross 4: ',CP4)
```

A resultante dos momentos vem assim:

\begin{equation*}
\begin{aligned}
\bullet\quad\sum \vec{M}_{o}
&= \vec{M}_O + (\vec{r}_{\textrm{OA}}\times\vec{R}_\textrm{A})
+ (\vec{r}_{\textrm{OB}}\times\vec{R}_\textrm{B}) = \vec{0}\\
&~\Leftrightarrow~(- 41569.2 + 200\,P)\,\vec{\imath} \\
&\quad + (100\,R_\textrm{Az} - 100\,R_\textrm{Bz})\,\vec{\jmath} \\
&\quad + (- 24000 -100\,R_\textrm{Ay} + 100\,R_\textrm{By})\,\vec{k} = \vec{0}
\end{aligned}
\end{equation*}

```python
expMo = CP1 + CP2 + CP3 + CP4
print('Exp Mo = ',expMo)
```

Com componentes escalares:

\begin{equation*}
\begin{cases}
\sum M_{ox} &=& 0 \\
\sum M_{oy} &=& 0 \\
\sum M_{oz} &=& 0
\end{cases}~\Leftrightarrow~
\begin{cases}
- 41569.2 + 200\,P &=& 0 \\
100\,R_\textrm{Az} - 100\,R_\textrm{Bz} &=& 0 \\
- 24000 -100\,R_\textrm{Ay} + 100\,R_\textrm{By} &=& 0
\end{cases}
\end{equation*}

**Equilíbrio dos Forças**

Contabilizando também a análise do equilíbrio estático pelo sumatório das forças aplicadas vem:

\begin{equation*}
\bullet\quad\sum \vec{F} = \vec{0}
~\Leftrightarrow~
\begin{cases}
\sum F_{x} =  0 \\
\sum F_{y} =  0 \\
\sum F_{z} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
R_\textrm{Ax} &=& 0 \\
R_\textrm{Ay} + R_\textrm{By} - P - T &=& 0 \\
R_\textrm{Az} + R_\textrm{Bz} &=& 0
\end{cases}
\end{equation*}

```python
EFX = rax
EFY = ray + rby - P - T
EFZ = raz + rbz
EMX = expMo.args[0][0].args[2][0]
EMY = expMo.args[0][0].args[2][1]
EMZ = expMo.args[0][0].args[2][2]

sol = solve({EFX,EFY,EFZ,EMX,EMY,EMZ},{rax,ray,raz,rby,rbz})

RAx = float(sol[rax]); print('RAx = ','{:.1f}'.format(RAx),' N')
RAy = float(sol[ray]); print('RAy = ','{:.1f}'.format(RAy),' N')
RAz = float(sol[raz]); print('RAz = ','{:.1f}'.format(RAz),' N')
RBy = float(sol[rby]); print('RBy = ','{:.1f}'.format(RBy),' N')
RBz = float(sol[rbz]); print('RBz = ','{:.1f}'.format(RBz),' N')
```
