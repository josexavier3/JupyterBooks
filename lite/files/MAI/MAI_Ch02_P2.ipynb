#### Questões de análise

### Resolução

Considere-se o diagrama de corpo livre da roldana em E (desprezando o atrito
e a massa):

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P2/MAI_Ch02_P2a.jpg"
style="max-width: 100%;"/>

Sobre a partícula E são exercidas as seguintes forças:

- (i) a força atribuida ao peso próprio do balde, devido à ação da força
gravítica,
- (ii) as forças de tração exercidas pelas duas extremidades do cabo onde
desliza a roldana.

A condição de equilíbrio estático das forças aplicas conduz ao sistema de
equações seguintes:

\begin{equation*}
\sum\vec{F} = \vec{0}
\end{equation*}

Escolhendo um sistema de coordenadas ortonormado, $S (0,x, y)$, pode-se
escrever as seguintes equações escalares:

\begin{equation*}
\begin{cases}
\sum F_x &= 0 \\[2ex]
\sum F_y &= 0
\end{cases}
\end{equation*}

**_i)_** Desenvolvendo a primeira equação de equilíbrio $(\sum F_x = 0)$ vem:

\begin{equation*}
\begin{aligned}
\sum F_x = 0  &\Leftrightarrow
T_1\sin\theta - T_2\sin\theta = 0 \\
&\Leftrightarrow T_1  = T_2 = T
\end{aligned}
\end{equation*}

De onde se conclui que as forças exercidas pelo cabo na roldana são iguais em
módulo. De notar que a condição de dimensionamento implica que:
$T_\textrm{max} = 5,8$ kN.

De notar que esta mesma conclusão poderia ter sido obtida pela lei dos Senos
(pelo método da linha poligonal das três forças aplicada na
partícula - a solução geométrica dá origem a um polígono fechado):

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P2/MAI_Ch02_P2b.jpg"
style="max-width: 100%;"/>

\begin{equation*}
\begin{aligned}
&\frac{T_1}{\sin\theta} = \frac{T_2}{\sin\theta} =
\frac{P}{\sin(\pi-2\theta)} \\
&\Rightarrow T_1 = T_2 = T_\textrm{max} = 5,8~\mathrm{[kN]}
\end{aligned}
\end{equation*}

**_ii)_** Da segunda equação de equilíbrio estático $(\sum F_y = 0)$ resulta:

\begin{equation*}
\begin{aligned}
\sum F_y = 0  & \Leftrightarrow
T_1\cos\theta + T_2\cos\theta - P = 0
\quad (\mathrm{com:}~T_1 = T_2 = T)\\
&\Leftrightarrow 2T \cos\theta = P \\
& \quad\therefore~ T = f (P, \theta) \\
\end{aligned}
\end{equation*}

Da expressão do equilíbrio se conclui que a força de tração do cabo ($T$) depende do peso ($P$) e do ângulo ($\theta$).

Uma vez que o valor de $P$ é fixo pelo peso próprio do balde (assume-se o seu valor invariável na solução, e eventualmente igual ao peso máximo carregado pelo balde para efeitos de dimensionamento), a força de tração, $T$, suportada pelo cabo será uma função do ângulo $\theta$, que a sua direção faz com o eixo $y$. Como é enunciado o valor admissível suportável pelo cabo, a equação anterior deve ser resolvida para o valor de ângulo máximo admissível pelo mecanismo:

\begin{equation*}
\begin{aligned}
&T_\textrm{max} \ge \frac{P}{2\cos\theta}~,
\quad \textrm{resolvendo para $\theta$ admissível:}\\
&\Leftrightarrow \cos\theta \ge \frac{P}{2T_\textrm{max}} \\
&\Leftrightarrow \theta \le \arccos\left(\frac{P}{2T_\textrm{max}}\right)~\textrm
{[radianos]}
\end{aligned}
\end{equation*}

```python
import numpy as np

Tmax = 5800. # unit: Newton, virgula flutuante
PesoE = 3000. # unit: N
costheta = PesoE/2/Tmax
print('cos(theta) = ','{:.3f}'.format(costheta),' [radianos]')
theta = np.arccos(costheta)
print('theta = ','{:.1f}'.format(np.rad2deg(theta)),' [graus]')
```

_Abordagem: lei dos senos_

De notar que a orientação dos cabos poderia ter sido obtida quer pela lei dos senos (LS):

\begin{equation*}
\begin{aligned}
\frac{T_\textrm{max}}{\sin\theta} =
\frac{P}{\sin(\pi-2\theta)}
\end{aligned}
\end{equation*}

Considerando relações trigonométricas,

\begin{equation*}
\begin{aligned}
\sin(\pi - 2\theta) = \sin(\pi)\cos(2\theta) - \cos(\pi)\sin(2\theta)
= \sin(2\theta) = 2\sin(2\theta)\cos(2\theta)
\end{aligned}
\end{equation*}

pelo que,

\begin{equation*}
\begin{aligned}
\frac{T_\textrm{max}}{\sin\theta} =
\frac{P}{2\sin(2\theta)\cos(2\theta)}
\Leftrightarrow
T_\textrm{max} = \frac{P}{2\cos(2\theta)}
\end{aligned}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import nsolve

thetaLCoS = sy.symbols('thetaLCoS')
EQLS = Tmax/sy.sin(thetaLCoS) - PesoE/sy.sin(np.pi - 2*thetaLCoS)
thetaLS = float(nsolve(EQLS, thetaLCoS, 1))
print('theta (Lei Senos) = ','{:.2f}'.format(np.rad2deg(thetaLS)),'[graus]')
```

_Abordagem: lei dos cossenos_

O mesmo cálculo poderia ter sido obtido pela lei dos Cosenos:

\begin{equation*}
\begin{aligned}
P^2 = T_1^2 + T_2^2 - 2T_1T_2\cos(\pi-2\theta)
\end{aligned}
\end{equation*}

Considerando relações trigonométricas,

\begin{equation*}
\begin{aligned}
\cos(\pi - 2\theta) &= \cos(\pi)\cos(2\theta) + \sin(\pi)\sin(2\theta)
= -\cos(2\theta) = -[\cos^2(\theta)-\sin^2(\theta)]\\
&= -\cos^2(\theta)+[1-\cos^2(\theta)]
= 1-2\cos^2(\theta)
\end{aligned}
\end{equation*}

de onde resulta ($T_1 = T_2 = T$),

\begin{equation*}
\begin{aligned}
P^2 &= 2T^2 - 2T^2[1-2\cos^2(\theta)] = 4T^2 \cos^2(\theta)\\
&\Leftrightarrow T^2 = \frac{P^2}{4\cos^2(\theta)}\\
&\therefore~~ T = \frac{P}{2\cos(\theta)}
\qquad\longrightarrow~~
\theta = \arccos\left(\frac{P}{2T} \right)
\end{aligned}
\end{equation*}

```python
print('code 1 -------------')
aux = np.arccos((PesoE**2 - 2*Tmax**2)/(-2*Tmax**2))
thetaC1 = (np.pi-aux)/2
print('theta (Lei Cosenos) = ','{:.2f}'.format(np.rad2deg(thetaC1)),'[graus]')

print('code 2 -------------')
EQ1 = PesoE**2 - 2*Tmax**2 + 2*Tmax**2*sy.cos(np.pi - 2*thetaLCoS)
thetaC2 = float(nsolve(EQ1, thetaLCoS, 1))
print('theta (Lei Cosenos) = ','{:.2f}'.format(np.rad2deg(thetaC2)),'[graus]')
```

**Altura máxima da estrutura**

A altura máxima $h$ será obtida considerando a relação geométrica do problema
(estrutura), assumindo o cálculo do ângulo admissível determinado anteriormente.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P2/MAI_Ch02_P2c.jpg"
style="max-width: 100%;"/>

Da geometria do problema resulta diretamente:

\begin{equation*}
\begin{aligned}
h_T = h + h_1(\theta)
\quad\Leftrightarrow\quad
h = h_T - h_1(\theta)
\end{aligned}
\end{equation*}

em que, $h_T$ é a altura total e $h_1$ é uma altura que pode ser relacionda com o ângulo $\theta$. Desta forma altura $h$ pode ser relacionada geometricamente com a altura toral e o ângulo $\theta$ (ver diagrama de corpo livre).

A altura total da escavação vem:

em que $h_T= 5+3(1.5) = 3.5 + 4(1.5)$ [m].

Por sua vez, $h_1$ vem:

\begin{equation*}
\begin{aligned}
\tan\theta = \frac{2.5}{h_1} \Leftrightarrow h_1
= \frac{2.5}{\tan\theta}
\end{aligned}
\end{equation*}

Pelo que,

\begin{equation*}
\begin{aligned}
h = 9.5 - \frac{2.5}{\tan\theta}
\end{aligned}
\end{equation*}

```python
comp = 5./2 # unit: m
hT = 5+3*1.5 # unit: m
h1 = comp/np.tan(theta)
h = hT - h1
print(f'theta = {np.rad2deg(theta):.2f} [deg]')
print('hT = ','{:.2f}'.format(hT),' [m]')
print('h1 = ','{:.2f}'.format(h1),' [m]')
print('h  = ','{:.2f}'.format(h),' [m]')
```

```python
import matplotlib.pylab as plt
from ipywidgets import interactive

t0 = np.round(np.rad2deg(np.arctan(comp/hT)))
print(f'Valor mínimo para o ângulo theta = {t0:.1f}, deg')
tcric = theta

def elevacao(ang):
    hatual = hT - comp/np.tan(np.deg2rad(ang))
    Tatual = PesoE/2/np.cos(np.deg2rad(ang))
    if Tatual >= Tmax:
        cor = 'r'
    else:
        cor = 'g'
    plt.figure(figsize=(6,4))
    # Cabo E
    caboE_x = [-comp,0] # x1, x2
    caboE_y = [hT,hatual] # y1, y2
    plt.plot(caboE_x, caboE_y,cor+'o-')
    # Cabo D
    caboD_x = [comp,0]
    caboD_y = [hT,hatual]
    plt.plot(caboD_x, caboD_y,cor+'s-')
    plt.text(-2,2,f'T={Tatual:.1f} N')
    plt.text(-2,1,f'h={hatual:.1f} m')
    # valores criticos
    plt.text(1.,2,f'T.max={Tmax:.1f} N')
    plt.text(1.,1.5,f'h.crit={h:.1f} m')
    plt.text(1.,1,f'ang.crit={np.rad2deg(theta):.1f} º')
    # Altura
    alt_x = [0, 0]
    alt_y = [0, hatual]
    plt.plot(alt_x, alt_y,cor+'v-')
    plt.show()

interactive_plot = interactive(elevacao, ang=(t0, 90))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
