#### Revisão

O coeficiente de atrito é uma quantidade adimensional que expressa a oposição que exercem as superfícies (rugosas) de dois corpos em contato ao deslizar um em relação ao outro.

O valor do coeficiente de atrito é caraterístico de cada par de materiais, e não sendo portanto uma propriedade intrínseca do material. Depende de muitos fatores tais como o acabamento das superfícies em contato, a velocidade relativa entre as superfícies, a temperatura, etc.

Geralmente distinguem-se dois valores:

- Coeficiente de atrito estático ($\mu_\textrm{e}$): é medido quando ambas as superfícies estão em repouso (sem se mover).

- Coeficiente de atrito cinético ($\mu_\textrm{c}$): é medido quando uma ou ambas as superfícies estão em movimento.

#### b) Determine a força mínima $\vec{F}$ que deverá estar aplicada para manter o sistema em equilíbrio.  Determine ainda a correspondente força de tração no cabo. <a id="B"></a> ([enunciatum](#PIV))

**Massa $\mathbf{m_2}$ :**

Do equilíbrio das forças envolvidas no diagrama de corpo livre da massa $m_2$ resulta:

\begin{equation*}
\vec{R}=\sum_{i}\vec{F}_i = 0
~\Leftrightarrow~
\begin{cases}
\sum  F_\textrm{x} = 0 \\
\sum F_\textrm{y} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
F - F_\textrm{N1} = 0 \\
-m_2g - F_\textrm{A1} + F_\textrm{T1} = 0
\end{cases}
\end{equation*}

com $F_\textrm{A1} = \mu_\textrm{e} F_\textrm{N1}$ e $\mu_\textrm{e} = 0.3$,

\begin{equation*}
\begin{cases}
\sum  F_\textrm{x} = 0 \\
\sum F_\textrm{y} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
F - F_\textrm{N1} = 0 \\
-m_2g - 0.3F_\textrm{N1} + F_\textrm{T1} = 0
\end{cases}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import solve

mu12 = 0.3
mu20 = 0.5

f, fn1, ft1 = sy.symbols('f fn1 ft1')

sumFx_2 = f - fn1
print('sumFx (m2) = ',sumFx_2)
sumFy_2 = -P2 - mu12*fn1 + ft1
print('sumFy (m2) = ',sumFy_2)
```

**Massa $\mathbf{m_1}$ :**

Do equilíbrio das forças envolvidas no diagrama de corpo livre da massa $m_1$ vem:

\begin{equation*}
\begin{cases}
\sum  F_\textrm{x} = 0 \\
\sum F_\textrm{y} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
F_\textrm{N1} - F_\textrm{N2} = 0 \\
-m_1g + F_\textrm{A1} + F_\textrm{A2} + F_\textrm{T2} = 0
\end{cases}
\end{equation*}

com $F_\textrm{A1} = 0.3 F_\textrm{N1}~(\mu_\textrm{e}=0.3)$, $F_\textrm{A2} =
0.5 F_\textrm{N2}~(\mu_\textrm{e}=0.5)$ e $F_\textrm{T2} = 2F_\textrm{T1}$
(do equilíbrio da roldana B),

\begin{equation*}
\begin{cases}
\sum  F_\textrm{x} = 0 \\
\sum F_\textrm{y} = 0
\end{cases}
~\Leftrightarrow~
\begin{cases}
F_\textrm{N1} - F_\textrm{N2}= 0 \\
-m_1g + 0.3F_\textrm{N1} + 0.5F_\textrm{N2} + 2F_\textrm{T1} = 0
\end{cases}
\end{equation*}

```python
fn2, ft2 = sy.symbols('fn2 ft2')

sumFx_1 = fn1 - fn2
print('sumFx (m1) = ',sumFx_1)
sumFy_1 = -P1 + mu12*fn1 + mu20*fn2 + 2*ft1
print('sumFy (m1) = ',sumFy_1)
```

**Sistema de equações**

Agrupando as equações de equilíbrio têm-se:

\begin{equation*}
\begin{cases}
F - F_\textrm{N1} = 0 \\
- 0.3F_\textrm{N1} + F_\textrm{T1} = m_2g \\
F_\textrm{N1} - F_\textrm{N2}= 0 \\
0.3F_\textrm{N1} + 0.5F_\textrm{N2} + 2F_\textrm{T1} = m_1g 
\end{cases}
\end{equation*}

```python
sol = solve({sumFx_2,sumFy_2,sumFx_1,sumFy_1},{f, fn1, ft1, fn2})
print('Sistema de equações --------------------')
F = sol[f]
print('F   = ','{:.1f}'.format(F),' N')
FN1 = sol[fn1]
print('FN1 = ','{:.1f}'.format(FN1),' N')
FN2 = sol[fn2]
print('FN2 = ','{:.1f}'.format(FN2),' N')
FT1 = sol[ft1]
print('FT1 = ','{:.1f}'.format(FT1),' N')

print('\nOutras variáveis --------------------')
FA1 = mu12*FN1
print('FA1 = ','{:.1f}'.format(FA1),' N')
FA2 = mu20*FN2
print('FA2 = ','{:.1f}'.format(FA2),' N')
FT2 = 2*FT1
print('FT2 = ','{:.1f}'.format(FT2),' N')
```

Numa resolução alternativa na forma matricial pode-se resolver:

\begin{equation*}
[A]\{x\}=\{b\} \Leftrightarrow
\begin{bmatrix}
1 & -1 & 0 & 0 \\
0 & -0.3 & 0 & 1 \\
0 & 0.3 & 0.5 & 2 \\
0 & 1 & 0 & -1
\end{bmatrix}
\begin{bmatrix}
F \\ F_\textrm{N1} \\ F_\textrm{N2} \\ F_\textrm{T1}
\end{bmatrix}=
\begin{bmatrix}
0 \\ m_2g \\ 0 \\ m_1g
\end{bmatrix}
\end{equation*}

cuja solução é: $\{x\}=[A]^{-1}\{b\}$

```python
import numpy as np
np. set_printoptions(precision=1, suppress=True)

A0 = [1, -1, 0, 0]
A1 = [0, -0.3, 0, 1]
A2 = [0, 1, -1, 0]
A3 = [0, 0.3, 0.5, 2]

A = np.array([A0,A1,A2,A3])
print(f'A =\n {A}')

invA = np.linalg.inv(A)
print(f'A^-1 =\n {invA}')

b = np.array([0,P2,0,P1])
print(f'b = {b}')
```

```python
print('Solução --------------------')
x = np.linalg.solve(A,b)
print(f'x = {x} [N]')
print(f'F = FN1 = FN2 = {x[0]:.1f} N')
print(f'FT1 = {x[3]:.1f} N')
print('+ outras forças ------------')
print(f'FA1 = {x[1]*mu12:.1f} N')
print(f'FA2 = {x[2]*mu20:.1f} N')
print(f'FT2 = {x[3]*2:.1f} N')
```
