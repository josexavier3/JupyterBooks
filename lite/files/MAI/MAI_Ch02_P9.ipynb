## **(a)** Sabendo que a força máxima que o cabo, que passa nas roldanas, pode suportar é de 30 N, determine a carga máxima Q que pode ser levantada e a força exercida sobre a placa em O

Comece-se por analisar o mecanismo de roldanas. Seccione-se o mecanismo em 3 partes distintas, representadas por 1, 2 e 3 na figura, isolando desta forma as três roldanas que serão analisadas como partículas sobre as quais atuam as forças externas. Em cada diagrama de corpo livre, identificam-se todas as forças externas aplicadas. Analisa-se de seguida o equilíbrio das forças na partícula.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P9/MAI_Ch02_P9a.jpg"
width="600">

A análise do equilíbrio estático nas secções 1, 2 e 3 permitem escrever o seguinte sistema de equações escalares (componente segundo o eixo vertical $y$):

\begin{equation*}
\begin{aligned}
\textrm{Secção 1:} \\
\textrm{Secção 2:} \\
\textrm{Secção 3:}
\end{aligned}
\Leftrightarrow
\begin{cases}
R_1 - 2F - R_2 &=& 0 \\
R_2 - 2F &=& 0 \\
3F - Q &=& 0
\end{cases}
\Leftrightarrow
\begin{cases}
R_1 &=& 4F\\
R_2 &=& 2F \\
Q &=& 3F
\end{cases}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
import numpy as np
import matplotlib.pyplot as plt

F = 30 # unit: N
r1, r2, q = sy.symbols('r1 r2 q')

EQA = r1 - 2*F - r2
EQB = r2 - 2*F
EQC = 3*F - q
sol = solve({EQA,EQB,EQC},{r1, r2, q})

R1 = float(sol[r1]); print('R1 = ','{:.1f}'.format(R1),' N')
R2 = float(sol[r2]); print('R2 = ','{:.1f}'.format(R2),' N')
Q = float(sol[q]);   print(' Q = ','{:.1f}'.format(Q),' N')
```

## **(b)** Determine qual deverá ser o comprimento mínimo para os cabos AB, AC e AD, sabendo que a força que cada cabo pode suportar não pode exceder 60 N. Considere $\alpha = 30^\circ$. Justifique.

De uma análise geométrica do problema pode-se estabelecer as seguintes relações geométricas para a determinação do comprimento dos cabos, $H$:


<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P9/MAI_Ch02_P9c.jpg" width="500"
/>

\begin{equation*}
\sin\theta = \frac{r}{H}
\Leftrightarrow
H = \frac{r}{\sin\theta}
\end{equation*}

Ou seja, para a determinação do comprimento $H$ é necessário conhecer o ângulo $\theta$. 

```python
# Define a range of theta values in degrees
theta_degrees = np.linspace(1, 90, 100) 

# Convert degrees to radians
theta_radians = np.deg2rad(theta_degrees)

# Calculate the corresponding H values for r = 1
r = 1
H = r / np.sin(theta_radians)

# Create the line plot
plt.figure(figsize=(8, 6))
plt.plot(theta_degrees, H, label='H = r / sin(θ)', color='b')
plt.xlabel('θ (degrees)')
plt.ylabel('H')
plt.title('Line Plot of H as a Function of θ for r = 1')
plt.grid(True)
plt.legend()
plt.show()
```

O valor de $\theta$ pode ser obtido por uma análise estático de equilíbrio do problema.

Considere-se a partícula no ponto A. Represente-se as forças externas aplicadas na partícula A, pelo seu diagrama de corpo livre (DCL).

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P9/MAI_Ch02_P9b.jpg" width="800"
/>

O equilíbrio estática permite escrever:

\begin{equation*}
\sum\vec{F}=\vec{0}
\Leftrightarrow
\vec{T}_\textrm{AB} + \vec{T}_\textrm{AC} + \vec{T}_\textrm{AD} +
\vec{R}_\textrm{1} = \vec{0}
\end{equation*}

De notar que o vetor $\vec{T}_\textrm{AD}$ está contido no plano $x-y$, pelo que terá apenas componentes cartesianas não nulas nessas direções.
Contudo, os vetores $\vec{T}_\textrm{AC}$ e $\vec{T}_\textrm{AB}$ têm uma orientação definda no espaço. Considere-se $\theta$, o ângulo que a linha de acção de $\vec{T}_\textrm{AC}$ faz com o eixo $y$. Resulta então:

\begin{equation*}
\vec{T}_\textrm{AC} = T_\textrm{AC}\sin\theta~\vec{\lambda}_\textrm{OC} -T_\textrm{AC}\cos\theta\vec~\jmath
\end{equation*}

em que por sua vez

\begin{equation*}
\vec{\lambda}_\textrm{OC} = -\cos\alpha~\vec\imath~ +\sin\alpha~\vec k
\end{equation*}

pelo que

\begin{equation*}
\vec{T}_\textrm{AC} = T_\textrm{AC}\sin\theta~(-\cos\alpha~\vec\imath~ +\sin\alpha~\vec k) -T_\textrm{AC}\cos\theta~\vec\jmath
\end{equation*}

Os vetores de força aplicadas no ponto A devem ser decompostos no sistema de coordenadas cartesiano escolhidos:

\begin{equation*}
\begin{cases}
\vec{T}_\textrm{AB} = (-T_\textrm{AB}\sin\theta\cos\alpha)\vec\imath +
(-T_\textrm{AB}\cos\theta)\vec\jmath +
(-T_\textrm{AB}\sin\theta\sin\alpha)\vec{k} \\
\vec{T}_\textrm{AC} = (-T_\textrm{AC}\sin\theta\cos\alpha)\vec\imath +
(-T_\textrm{AC}\cos\theta)\vec\jmath +
(T_\textrm{AC}\sin\theta\sin\alpha)\vec{k} \\
\vec{T}_\textrm{AD} = (T_\textrm{AD}\sin\theta)\vec\imath +
(-T_\textrm{AD}\cos\theta)\vec\jmath \\
\vec{R_1} = R_1 \vec{\jmath}
\end{cases}
\end{equation*}

O sistema de equações de equilíbrio resulta portanto:

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0 \\
\sum F_z = 0
\end{cases}\Leftrightarrow
\begin{cases}
-T_\textrm{AB}\sin\theta\cos\alpha
-T_\textrm{AC}\sin\theta\cos\alpha
+ T_\textrm{AD}\sin\theta = 0 \\
-T_\textrm{AB}\cos\theta - T_\textrm{AC}\cos\theta
-T_\textrm{AD}\cos\theta + R_1 = 0\\
-T_\textrm{AB}\sin\theta\sin\alpha + T_\textrm{AC}\sin\theta\sin\alpha
= 0
\end{cases}
\end{equation*}

O sistema contem 3 equações e apresenta 4 incógnitas ($T_\textrm{AB}$, $T_\textrm{AC}$,  $T_\textrm{AD}$ e $\theta$). Pegando na terceira equação do sistema, conduto, conclui-se imediatamente:

\begin{equation*}
T_\textrm{AB} = T_\textrm{AC} \equiv T
\end{equation*}

Note-se que a mesma conclusão poderia ter sido encontrada geometricamente considerando a simetria do  disco em relação ao eixo $x$.

O sistema de equações fica asssim reduzido a 2 equações a 3 incógnitas:

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0 \\
\end{cases}\Leftrightarrow
\begin{cases}
-2T\sin\theta\cos\alpha
+ T_\textrm{AD}\sin\theta = 0 \\
-2T\cos\theta -T_\textrm{AD}\cos\theta + R_1 = 0\\
\end{cases}
\end{equation*}

Considerando $\alpha = 30^\circ$ (do enunciado) a primeira expressão ($\sum F_x = 0$)
simplifica-se:

\begin{equation*}
T_\textrm{AD} = 2T\cos(30^\circ\frac{\pi}{180^\circ}) = \sqrt{3}\,T
\end{equation*}

ou seja, conclui-se: 

\begin{equation*}
T_\textrm{AD} > T \qquad \textrm{(com  $T \equiv  T_\textrm{AB} =
T_\textrm{AC}$)}
\end{equation*}

Como as forças nos cabos não podem ultrapassar os 60 N (do enunciado):

\begin{equation*}
T_\textrm{AD} = T_\textrm{max} = 60~\textrm{N}
\quad\therefore\quad 
T = \frac{T_\textrm{AD}}{\sqrt{3}} =  \frac{60}{\sqrt{3}} = 34.6~\textrm{N}
\end{equation*}

Da segunda expressão de equilíbrio ($\sum F_y = 0$) vem finalmente:

\begin{equation*}
\cos\theta = \frac{R_1}{2T + T_\textrm{AD}} 
\quad\therefore\quad \theta = 21,8^\circ
\end{equation*}

```python
tab, tac, angT = sy.symbols('tab tac angT')
alpha1 = np.deg2rad(30) # unit: radianos

TAD = 60 # unit: N
print('TAD = ','{:.1f}'.format(TAD),' N')

# note: tab - tac
EQX = -tab*sy.sin(angT)*np.cos(alpha1) \
      -tac*sy.sin(angT)*np.cos(alpha1) \
      +TAD*sy.sin(angT)
EQY = -tab*sy.cos(angT) - tac*sy.cos(angT) - TAD*sy.cos(angT) + R1
EQZ = -tab*sy.sin(angT)*np.sin(alpha1) + tac*sy.sin(angT)*np.sin(alpha1)
sol = solve([EQX,EQY,EQZ],[tab, tac, angT])
# sol = nsolve([EQX,EQY,EQZ],[tab, tac, angT],[30,30,.5])

TAB = float(sol[2][0]); print('TAB = ','{:.1f}'.format(TAB),' N')
TAC = float(sol[2][1]); print('TAB = ','{:.1f}'.format(TAC),' N')
AngTheta = float(sol[2][2])
AngThetaDeg = np.rad2deg(AngTheta)
print('angTheta = ','{:.1f}'.format(AngThetaDeg),' graus')
```

Finalmente, o comprimento dos cabos é determinado pela relação geométrica:

\begin{equation*}
H = \frac{r}{\sin\theta}
\end{equation*}

```python
raio = 250 # unit: mm
H = raio/np.sin(AngTheta)
print('H = ','{:.1f}'.format(H),' mmm')
```

## **(c)** Em que condições a força de tração nos três cabos é igual? Justifique.

A condição de força de tração igual nos três cabos escreve-se:

\begin{equation*}
T_\textrm{AB} = T_\textrm{AC} = T_\textrm{AD} = T
\end{equation*}

Da expressão de equílbrio $\sum F_x = 0$, resulta:

\begin{equation*}
\begin{aligned}
&~-T\sin\theta\cos\alpha
-T\sin\theta\cos\alpha
+ T\sin\theta = 0 \\
\Leftrightarrow~&~
T\sin\theta (-2\cos\alpha + 1) = 0 \\
\Leftrightarrow~&~
\cos\alpha = \frac{1}{2} \\
\Rightarrow~&~
\alpha = 60^\circ
\end{aligned}
\end{equation*}

```python
alphaC = np.rad2deg(np.arccos(0.5))
print('alpha (c) = ','{:.1f}'.format(alphaC),' graus')
```
