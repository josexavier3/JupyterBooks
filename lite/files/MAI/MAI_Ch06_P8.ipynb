### Resolução

#### a) Determine o momento polar de inércia de superfície em relação à origem do referencial. <a id="A"></a> ([_enunciatum_](#P8))

Para determinar o momento polar em relação à origem do referencial $S(0, \vec{\imath},\vec{\jmath})$ pode-se calcular em primeiro lugar, os momentos de 2ª ordem da secção, ou momentos de inércia, relativamente aos eixos $x$ e $y$.

Para isso, considera-se a decomposição da figura conforme se mostra na figura abaixo, obtida pela composição de um retângulo subtraindo-se dois semi-círculos:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P8/MAI_Ch06_P8a.jpg"
width="1000">

O momento polar de inércia da superfície em relação à origem pode ser calculado por:

\begin{equation*}
J_0 = I_x + I_y
\quad\wedge\quad
I_x = I_x^{[1]} - I_x^{[2]} - I_x^{[3]}
\quad\wedge\quad
I_y = I_y^{[1]} - I_y^{[2]} - I_y^{[3]}
\end{equation*}

ou seja,

\begin{equation*}
\begin{aligned}
I_i = I_i^\Box - I_i^\textrm{semi-circulo,Esq}
- I_i^\textrm{semi-circulo,Dir} \quad (i = x,y) \\
\end{aligned}
\end{equation*}



**Área [1] - Retângulo**

Da leitura direta do formulário da tabela é possível retirar expressões em relação ao sistema de coordenadas $S(C,x',y')$. Contudo é necessário confrontar esse referencial com o referencial escolhido para os cálculos do problem. De notar que os eixos $x'$ e $x$ são coincidentes, contudo o eixo $y'$ e $y$ são colineares (e.g. têm a mesma direção) mas afastados de uma certa distância. Neste último caso, será portanto necessário recorrer ao teorema dos eixos paralelos para expressar o momento de 2ª ordem de superfície em relação ao eixo global escolhido.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P8/MAI_Ch06_P8b.jpg"
width="1000">

\begin{equation*}
\begin{aligned}
\textrm{Referencial}~S(C,x',y'):\\
I_{x'} &= \frac{bh^3}{12} = \frac{(3a)(2a)^3}{12} = 2a^4\\
I_{y'} &= \frac{bh^3}{12} = \frac{(2a)(3a)^3}{12} =
\frac{9a^4}{2} \\
\textrm{Referencial}~S(O,x,y):\\
I_{x} &= I_{x'} = 2a^4\\
I_{y} &= I_{y'} + A_1 d_1^2\quad\textrm{(teorema dos eixos paralelos)} \\
&= \frac{9a^4}{2} + (2a)(3a)(a/2)^2 = 6a^4
\end{aligned}
\end{equation*}

```python
import numpy as np
import sympy as sy

def printsymSF(expr):
    expr2 = sy.simplify(sy.factor(expr))
    return expr2.xreplace({n : round(n, 3) for n in expr.atoms(sy.Number)})

a = sy.symbols('a') # unit: mm
print('---------------------')
print('Área [1] - Retângulo')
print('---------------------')
b, h = 3*a, 2*a
print(f'b = {b} mm | h = {h} mm')
IxC1 = b*h**3/12
print(f'Ix (local centroid - table) = {IxC1} mm4')
IyC1 = h*b**3/12
print(f'Iy (local centroid - table) = {IyC1} mm4')
d1, A1 = a/2, b*h
print(f'd1 = {d1} mm | A1 = {A1} mm2')
IxSC1 = IxC1
print(f'Ix (global centroid) = {IxSC1} mm4')
IySC1 = IyC1 + A1*d1**2
print(f'Iy (global centroid, Parallel Axis Theorem) = {IySC1} mm4')
```

**Área [2] - semi-círculo (direito)**

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P8/MAI_Ch06_P8c.jpg"
width="1000">

\begin{equation*}
\begin{aligned}
\textrm{Referencial}~S(C,x',y'):\\
I_{x'} &= \frac{\pi r^4}{8} = \frac{\pi a^4}{8}
= 0.393 \,a^4\\
I_{y''}\Bigg|_{\textrm{(dado na tabela)}} &= I_{y'}\Bigg|_{\textrm{(eixo
passa no centroide $C$)}} + Ad^2
\quad\wedge\quad d = 4r/3\pi|_{r=a}\\
I_{y'}&=
\frac{\pi a^4}{8} - \left(\frac{\pi a^2}{2}\right)
\left(\frac{4a}{3 \pi}\right)^2 \\
&= \left(-\frac{8}{9\pi} +  \frac{\pi}{8} \right)a^4
= 0.110 \,a^4
\\
\textrm{Referencial}~S(O,x,y):\\
I_{x} &= I_{x'} = \frac{\pi a^4}{8}\\
I_{y} &= I_{y'} + A_2 d_2^2 \quad\textrm{(teorema dos eixos paralelos)}\\
&= \left(-\frac{8}{9\pi}
+ \frac{\pi}{8} \right)a^4
+ \left(\frac{\pi a^2}{2}\right)
(\underbrace{2a-\frac{4a}{3\pi}}_{d_2 - \textrm{ver fig.}})^2 \\
& = \left(-\frac{8}{3}
+ \frac{17\pi}{8} \right)a^4 = 4.01\,a^4
\end{aligned}
\end{equation*}

```python
print('---------------------------------')
print('Área [2] - semi-círculo (direito)')
print('---------------------------------')
IxC2 = np.pi*a**4/8
print(f'Ix (local centroid - table) = {printsymSF(IxC2)} mm4')
A2 = np.pi*a**2/2
print(f'A2 (table) = {printsymSF(A2)} mm2')
d12 = 4*a/3/np.pi
print(f'd12 = {printsymSF(d12)} mm')
IyC2 = np.pi*a**4/8 - A2*d12**2
print(f'Iy (local centroid - table) = {printsymSF(IyC2)} mm4')
d2 = 2*a - 4*a/3/np.pi
print(f'd1 = {printsymSF(d2)} mm')
IxSC2 = IxC2
print(f'Ix (global centroid) = {printsymSF(IxSC2)} mm4')
IySC2 = IyC2 + A2*d2**2
print(f'Iy (global centroid) = {printsymSF(IySC2)} mm4')
```

**Área [3] - semi-círculo (esquerdo)**

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P8/MAI_Ch06_P8d.jpg"
width="400">

\begin{equation*}
\begin{aligned}
\textrm{Referencial}~S(O,x',y'):\\
I_{x'} &= \frac{\pi r^4}{8} = \frac{\pi a^4}{8}
= 0.393 a^4\\
I_{y'}\Big|_\textrm{teorema dos eixos paralelos} &=
\frac{\pi a^4}{8} - \left(\frac{\pi a^2}{2}\right)
\left(\frac{4a}{3 \pi}\right)^2 \\
&= \left(-\frac{8}{9\pi} +  \frac{\pi}{8} \right)a^4
= 0.110 \,a^4
\\
\textrm{Referencial}~S(O,x,y):\\
I_{x} &= I_{x'} = \frac{\pi a^4}{8}\\
I_{y} &= I_{y'} + A_3 d_3^2\\
&= \left(-\frac{8}{9\pi}
+ \frac{\pi}{8} \right)a^4
+ \left(\frac{\pi a^2}{2}\right)
(\underbrace{a-\frac{4a}{3\pi}}_{d_3 \textrm{(fig.)}})^2 \\
& = \left(-\frac{4}{3}
+ \frac{5\pi}{8} \right)a^4 = 0.63\,a^4
\end{aligned}
\end{equation*}

```python
print('---------------------------------')
print('Área [2] - semi-círculo (esquerdo)')
print('---------------------------------')
IxC3 = np.pi*a**4/8
print('Ix (local centroid - table) = ',IxC3,' mm4')
A3 = np.pi*a**2/2
print('A3 (table) = ',A3,' mm2')
d13 = 4*a/3/np.pi
print('d12 = ',d12,' mm')
IyC3 = np.pi*a**4/8 - A3*d13**2
print('Iy (local centroid - table) = ',IyC3,' mm4')
d3 = a - 4*a/3/np.pi; print('d1 = ',d3,' mm')
IxSC3 = IxC3
print('Ix (global centroid) = ',IxSC3,' mm4')
IySC3 = IyC3 + A3*d3**2
print('Iy (global centroid) = ',IySC3,' mm4')
```

\begin{equation*}
\begin{aligned}
\textrm{Referencial}~S(O,x,y):\\
\bullet\quad I_x &= I_x^{[1]} - I_x^{[2]} - I_x^{[3]} \\
I_{x} &= 2a^4 - \frac{\pi a^4}{8} - \frac{\pi a^4}{8}\\
&= \left( 2 - \frac{\pi}{4} \right) a^4 = 1.215 \,a^4\\
\bullet\quad I_y &= I_y^{[1]} - I_y^{[2]} - I_y^{[3]}\\
I_{y} &= 6a^4
-  \left(-\frac{4}{3}
+ \frac{5\pi}{8} \right)a^4 -
\left(-\frac{8}{3}
+ \frac{17\pi}{8} \right)a^4 \\
& = \left( 10 - \frac{11\pi}{4} \right) a^4
= 1.361 \, a^4\\
\bullet\quad J_0 &= I_x + I_y = \left( 2 - \frac{\pi}{4} \right) a^4
+ \left( 10 - \frac{11\pi}{4} \right) a^4 \\
&= (12  - 3 \pi)\,a^4 = 2.575 \, a^4
\end{aligned}
\end{equation*}

```python
IX = IxSC1 - IxSC2 - IxSC3
print(f'Ix = Ix1 - Ix2 - Ix3 = {printsymSF(IX)} mm4')

IY = IySC1 - IySC2 - IySC3
print(f'Iy = Iy1 - Iy2 - Iy3 = {printsymSF(IY)} mm4')

J0 = IX + IY
print(f'J0 = Ix + Iy         = {printsymSF(J0)} mm4')
```

Considerando $a= 0.02$ m calcula-se:

\begin{equation*}
J_0 = 2.575 \, (0.02)^4 = 4.120\times 10^{-7}
~\textrm{m$^4$}
\end{equation*}

```python
A = 0.02 # unit: m
J0val = J0.subs(a,A)
print(f'J0 (a = 0.02 m) = {J0val:6.4e} m4')
```

#### b) Diga o que entende por eixos principais de inércia de uma superfície. Justifique se os eixos $x$ e $y$ do referencial representado são eixos principais de inércia. <a id="B"></a> ([_enunciatum_](#P8))

Os eixos principais de inércia definem um sistema de coordenadas ortonormal e direto em relação ao qual o produto de inércia é nulo:

\begin{equation*}
I_{xy} = \int xy \mathrm{d}A= 0
\end{equation*}

Nesse sistema de coordenadas, os momentos de 2º ordem de superfície são momentos principais com valores máximos e mínimos.

Se for possível definir pelo menos um eixo de simetria da superfície em análise esse eixo é um eixo principal e o produto de inércia ($I_{xy}$) é nulo em relação a este.

Como o eixo $x$ é um eixo de simetria, então o produto de inércia $I_{xy}$ é nulo e o sistema de eixos $S(0,\vec{\imath},\vec{\jmath})$ conduzem a momentos de 2ª ordem máximos e mínimos, respetivamente.


#### c) Considere agora uma placa fina no referencial $S(0,x,y,z)$ com a superfície a sombreado, com espessura constante ($t=5$ mm) e massa específica $\rho=$ 7850 kg/m $^3$. Determine o momento de inércia da placa em relação ao eixo $z, I_z$. <a id="C"></a> ([_enunciatum_](#P8))

Para uma placa delgada de espessura uniforme $t$, feita de um material homogêneo de massa específica $\rho$, o momento de inércia em relação a um eixo AA’ contido no plano da placa é:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P8/MAI_Ch06_P8e.jpg"
width="1000">

\begin{equation*}
J_{AA', \textrm{massa}} = \int r^2 \mathrm{d}m \Bigg|_{\mathrm{d}m = \rho t \mathrm{d}A}
= \rho t  \int r^2 \mathrm{d}A
= \rho t  I_{AA', \textrm{area}}
\end{equation*}

De modo semelhante, para um eixo BB’ contido no plano da placa e perpendicular a AA’, tem-se

\begin{equation*}
J_{BB', \textrm{massa}}
= \rho t  \int r^2 \mathrm{d}A
= \rho t  I_{BB', \textrm{area}}
\end{equation*}

Considerando agora o eixo CC’ perpendicular ao plano de placa e passando pelo ponto de intersecção C de AA’ e BB’, escrevemos:

\begin{equation*}
J_{CC', \textrm{massa}}
= \rho t  \underbrace{I_{C, \textrm{area}}}_{J_C}
= \rho t  (I_{AA', \textrm{area}} + I_{BB', \textrm{area}})
\end{equation*}

Aplicando a expressão anterior ao problema, o momento de segunda ordem de área polar da placa em relação ao eixo $z$ é:

\begin{equation*}
J_z = \rho t (I_{x, \textrm{area}} + I_{y, \textrm{area}}) = \rho t J_0
\end{equation*}

desenvolvendo,

\begin{equation*}
J_z = \rho t (12  - 3 \pi)\,a^4
= 101.1\,a^4
\end{equation*}

```python
rho = 7850
t = 5e-3 # unit: m
Jz = rho*t*J0
print(f'Jz = rho*t*J0 = {printsymSF(Jz)} m4')
```

Considerando $a= 0.02$ m calcula-se:

\begin{equation*}
J_z = 101.1\,(0.02)^4 = 1.617\times 10^{-5}~\textrm{m$^4$}
\end{equation*}

```python
Jzval = rho*t*J0val
print(f'Jz = {Jzval:6.4e} m4')
```
