#### Problema 3 <a name="PIII"></a>

A barra AB repousa sobre duas superfícies inclinadas, como se mostra na figura. A barra tem comprimento $L$ e suporta uma carga concentrada $P$ aplicada à distância $a$ da sua extremidade A.  Sabe-se que o coeficiente de atrito estático entre a barra e as superfícies é 0,5. Determine qual deverá ser a distância mínima, $a$, para que a barra não escorregue sobre as superfícies.

#### Resolução

Comece-se por considerar o diagrama de corpo livre da barra. Nos pontos de ligação da barra às superfícies inclinadas surgem forças de reação cujas componentes podem ser desenvolvidas num referencial local com a orientação dos planos das superfícies. Para se determinar os sentidos (físicos) das forças de atrito geradas nos Pontos $A$ e $B$, no equilíbrio da força aplicada $P$,
note-se que ao aproximar-se a força $P$ do Ponto $A$ vai-se suscitar a rotação da viga no sentido anti-horário, conforme se ilustra na
Fig. 4(a).  De forma a contrariar essa rotação, as forças de atrito entre a viga e a superfície de apoio devem opor-se a esse movimento. Por conseguinte faz-se o DCL do problema como representado na Fig. 4(b).


<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P3/MAI_Ch03_P3a.jpg"
width="900">

A análise do equilíbrio estático do corpo da barra permite escrever (equações de equilibrio no plano):

\begin{equation*}
\begin{aligned}
\vec{R} = \sum_i\vec{F}_i = 0
\qquad\wedge\qquad
\vec{M}_\textrm{0}^\textrm{R} = \sum_i\vec{M}_\textrm{0}
= \sum_i(\vec{r}_\textrm{0i}\times\vec{F}_i) = 0
\end{aligned}
\end{equation*}

Do DCL comece-se por escrever os vetores forças de reações nos pontos $A$ e $B$:

\begin{equation*}
\begin{cases}
\vec{F}_\textrm{NA} &=& F_\textrm{NA}\cos(30)&\vec{\imath}
&+& F_\textrm{NA}\sin(30)&\vec{\jmath} \\
\vec{F}_\textrm{aA} &=& - F_\textrm{aA}\cos(60)&\vec{\imath}
&+& F_\textrm{aA}\sin(60)&\vec{\jmath}\\[1ex]
\vec{F}_\textrm{NB} &=& -F_\textrm{NB}\cos(60)\,&\vec{\imath}
&+& F_\textrm{NB}\sin(60)&\vec{\jmath}\\
\vec{F}_\textrm{aB} &=& - F_\textrm{aB}\cos(30)\,&\vec{\imath}
&-& F_\textrm{aB}\sin(30)&\vec{\jmath}\\
\end{cases}
\end{equation*}

Por sua vez, a resultante das forças de reação nos pontos $A$ e $A$ vêm:

| ang | 30 $^\circ$ | 60 $^\circ$ |
|:--|:--|:--|
| $\cos$ | $\frac{\sqrt{3}}{2}$ | $\frac{1}{2}$ |
| $\sin$ | $\frac{1}{2}$ | $\frac{\sqrt{3}}{2}$ |

\begin{equation*}
\begin{aligned}
\vec{F}_\textrm{A} &= \vec{F}_\textrm{NA} + \vec{F}_\textrm{aA} \\
&=[F_\textrm{NA}\cos(30)- F_\textrm{aA}\cos(60)]\,\vec{\imath}
+[F_\textrm{NA}\sin(30)+F_\textrm{aA}\sin(60)]\,\vec{\jmath}\\
&=\left(\frac{\sqrt{3}}{2}F_\textrm{NA} - \frac{1}{2}F_\textrm{aA}\right)\,
\vec{\imath}
+\left(\frac{1}{2}F_\textrm{NA} + \frac{\sqrt{3}}{2}F_\textrm{aA}\right)\,
\vec{\jmath}
\end{aligned}
\end{equation*}

\begin{equation*}
\begin{aligned}
\vec{F}_\textrm{B} &= \vec{F}_\textrm{NB} + \vec{F}_\textrm{aB} \\
&=[-F_\textrm{NB}\cos(60)- F_\textrm{aB}\cos(30)]\,\vec{\imath}
+[F_\textrm{NB}\sin(60)-F_\textrm{aB}\sin(30)]\,\vec{\jmath}\\
&=\left(-\frac{1}{2}F_\textrm{NB} - \frac{\sqrt{3}}{2}F_\textrm{aB}\right)\,
\vec{\imath}
+\left(\frac{\sqrt{3}}{2}F_\textrm{NB} - \frac{1}{2}F_\textrm{aB}\right)\,
\vec{\jmath}
\end{aligned}
\end{equation*}

A força de atrito estático máxima ($F_\textrm{m}$) e a força normal (de reação) ($F_\textrm{N}$) estão relacionadas pelo coeficiente de atrito estático:

\begin{equation*}
F_\textrm{m} = \mu_e F_\textrm{N}
\end{equation*}

Esta equação é válida no cenário de movimento iminente da barra sobre as superfícies inclinadas. Neste caso, assume-se o instante para o qual ($\mu_e = 0.5$):

\begin{equation*}
F_\textrm{aA} = 0.5  F_\textrm{NA}
\quad\textrm{e}\quad
F_\textrm{aB} = 0.5  F_\textrm{NB}
\end{equation*}

Assim sendo resultam as equações de equilíbrio estátio da resultantes das forças:

\begin{equation*}
\begin{aligned}
\vec{F}_\textrm{A}
&=\left(\frac{\sqrt{3}}{2} - \frac{1}{2}(0.5)\right)F_\textrm{NA}\,
\vec{\imath}
+\left(\frac{1}{2} + \frac{\sqrt{3}}{2}(0.5)\right)F_\textrm{NA}\,
\vec{\jmath}\\
&=\left(\frac{2\sqrt{3}-1}{4}\right)F_\textrm{NA}\,
\vec{\imath}
+\left(\frac{\sqrt{3}+2}{4}\right)F_\textrm{NA}\,
\vec{\jmath}\\
&=F_\textrm{NA}(0.616\,\vec{\imath}+0.933\,\vec{\jmath})
\end{aligned}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
from sympy.physics.mechanics import ReferenceFrame

fna = sy.symbols('fna')

mu = 0.5

ang60 = np.deg2rad(60) # unit: radianos
ang30 = np.deg2rad(30) # unit: radianos

faa = mu*fna
S = ReferenceFrame('S') # ||x|
vecFA = (fna*np.cos(ang30) - faa*np.cos(ang60))*S.x \
        + (fna*np.sin(ang30) + faa*np.sin(ang60))*S.y
print('vecFA : ',vecFA)
```

\begin{equation*}
\begin{aligned}
\vec{F}_\textrm{B}
&=\left(-\frac{1}{2} - \frac{\sqrt{3}}{2}(0.5)\right)F_\textrm{NB}\,
\vec{\imath}
+\left(\frac{\sqrt{3}}{2} - \frac{1}{2}(0.5)\right)F_\textrm{NB}\,
\vec{\jmath}\\
&=\left(- \frac{\sqrt{3}+2}{4}\right)F_\textrm{NB}\,
\vec{\imath}
+\left(\frac{2\sqrt{3}-1}{4}\right)F_\textrm{NB}\,
\vec{\jmath}\\
&=F_\textrm{NB}\,(-0.933\,\vec{\imath}
+0.616\,
\vec{\jmath})
\end{aligned}
\end{equation*}

```python
fnb = sy.symbols('fnb')
fab = mu*fnb
vecFB = (-fnb*np.cos(ang60) - fab*np.cos(ang30))*S.x \
        + (fnb*np.sin(ang60) - fab*np.sin(ang30))*S.y
print('vecFB : ',vecFB)
```

**Sistema de equações do equilíbrio estático**

Determinadas as componentes cartesianas das forças que atuam nos pontos $A$ e $B$ é expecito a desenvolvimento do sistema de equações que estabelecem o equilíbrio:

_Sumatório das forças em $x$ igual a zero_

\begin{equation*}
\rightarrow(+)~\sum F_\textrm{x} = 0~\Leftrightarrow~
\left(\frac{2\sqrt{3}-1}{4}\right)F_\textrm{NA} - \left(
\frac{\sqrt{3}+2}{4}\right)F_\textrm{NB}
= 0
\end{equation*}

```python
vecFA_X = vecFA.args[0][0].args[2][0]
vecFB_X = vecFB.args[0][0].args[2][0]
sumF_X = vecFA_X + vecFB_X
print(f'sumFX = {sumF_X} = 0')
```

_Sumatório das forças em $y$ igual a zero_

\begin{equation*}
\uparrow(+)~\sum F_\textrm{y} = 0~\Leftrightarrow~
\left(\frac{\sqrt{3}+2}{4}\right)F_\textrm{NA}
+ \left(\frac{2\sqrt{3}-1}{4}\right)F_\textrm{NB} - P
= 0
\end{equation*}

```python
p = sy.symbols('p')

vecFA_Y = vecFA.args[0][0].args[2][1]
vecFB_Y = vecFB.args[0][0].args[2][1]
sumF_Y = vecFA_Y + vecFB_Y - p
print(f'sumFY = {sumF_Y} = 0')
```

#### Somatório de momentos relativamente a $A$

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P3/MAI_Ch03_P3b.jpg"
width="900">

\begin{equation*}
\circlearrowleft(+)~:\sum M_\textrm{A}= 0~\Leftrightarrow~
\end{equation*}

\begin{equation*}
\begin{aligned}
-Pa - F_\textrm{a$_B$}\sin(30{\small\frac{\pi}{180}})L+
F_\textrm{N$_B$}\cos(30{\small\frac{\pi}{180}})L &= 0\\
-Pa + \left(- (0.5)\frac{1}{2}+\frac{\sqrt{3}}{2}
\right)LF_\textrm{N$_B$} &= 0\\
-Pa + \left(\frac{2\sqrt{3}-1}{4}
\right)LF_\textrm{N$_B$} &= 0\\
\end{aligned}
\end{equation*}

Finalmente juntando as equações resulta o seguinte sistema:

\begin{equation*}
\begin{cases}
\displaystyle\left(\frac{2\sqrt{3}-1}{4}\right)F_\textrm{NA} - \left(
\frac{\sqrt{3}+2}{4}\right)F_\textrm{NB}
&=& 0 \\[1ex]
\displaystyle\left(\frac{\sqrt{3}+2}{4}\right)F_\textrm{NA}
+ \left(\frac{2\sqrt{3}-1}{4}\right)F_\textrm{NB} - P
&=& 0\\[1ex]
-Pa + \displaystyle\left(\frac{2\sqrt{3}-1}{4}
\right)LF_\textrm{N$_B$} &=& 0\\
\end{cases}
\end{equation*}

```python
a, L = sy.symbols('a L')

EMA = -p*a + (-mu*np.sin(ang30) + np.cos(ang30))*L*fnb

sol = solve({sumF_X,sumF_Y,EMA},{a,p,fna})
print('-------------------')
print('Solução')
print('-------------------')
param = sol[0]
print('param : ',param)
print('a = ',param[a],' m')
print('fna = ',param[fna],' m')
print('P = ',param[p],' N')
print('\n----------------------------------------------------')
print('Expressão das forças de reação em função da força P')
print('----------------------------------------------------')
print(f'NA = {param[fna].args[0]/param[p].args[0]:.3f} P [N]')
print(f'FaA (= 0.5 NA) = {mu*param[fna].args[0]/param[p].args[0]:.3f} P [N]')

print(f'NB = {1/param[p].args[0]:.3f} P [N]')
print(f'FaB (= 0.5 NB) = {mu/param[p].args[0]:.3f} P [N]')
```

## Apêndice

Para explicar a relação entre o coeficiente de atrito estático ($\mu_e$) e o ângulo de atrito estático ($\phi_e$), comece-se por considerar um bloco de peso $W$, simplesmente assente numa superfície rugosa, como representado na Fig. 1.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P3/MAI_Ch03_P3i.jpg"
width="900">


De acordo com o diagrama de corpo livre, as únicas forças que atuam no bloco são o peso $W$ e a reação normal da superfície $N$ (Fig. 1a). Suponha-se, num  segundo momento, que uma força horizontal $P$ é aplicada ao corpo como se mostra na Fig. 1(b). Se a força $P$ for relativamente pequena, temos a experiência de que o bloco não se deslocará. Esta observação sugere que terá de existir uma força horizontal tal que o equilíbrio estático se verifique. Esta força iguala a força $P$ em magnitude e sentido oposto à força $P$ impedindo que o movimento de translação do bloco. Essa força, representa-se por $F$ e é designada por força de atrito estático.

Se a força aplicada $P$ for gradualmente aumentada (Fig. 1c), a força de atrito $F$ também aumentará continuando a opor-se a $P$ até que a sua intensidade atinja um valor máximo admissível $F_m$ (Fig. 1c). Esse valor máximo é dado por:

\begin{equation*}
F_m = \mu_e \cdot N
\end{equation*}

em que $\mu_e$ é o coeficiente de atrito estático.

Contudo, se a força aplicada $P$ continuar a aumentar, reconhece-se da observação experimental, que a força de atrito não conseguirá continuar a equilibrá-la, e o bloco começará a escorregar. Assim que o bloco inicia o movimento, a intensidade da força de atrito $F$ cai de $F_m$  para um valor inferior $F_c$ (Fig 1 c), vulgarmente designado por força de atrito cinético. Por sua vez, esse valor é por:

\begin{equation*}
F_c = \mu_c \cdot N
\end{equation*}

em que $\mu_c$ é o coeficiente de atrito cinético. Os valores de $\mu_e$ e $\mu_c$ não dependem da área das superfícies em contacto. Contudo, dependem fortemente da natureza das superfícies em contacto (Tabela 1).

Valores Aproximados de Coeficiente de atrito estático para Superfícies não lubrificadas

_Valores de coeficiente de atrito estático para superfícies não lubrificadas_

|Superfícies secas|$\mu_e$|
|:---------|:------------|
|Metal sobre metal | 0,15-0,60 |
|Metal na madeira | 0,20-0,60 |
|Metal sobre pedra | 0,30-0,70 |
|Metal em couro | 0,30-0,60 |
|Madeira sobre madeira | 0,25-0,50 |
|Madeira sobre couro | 0,25-0,50 |
|Pedra sobre pedra | 0,40-0,70|

Identificam-se assim, 4 (quatro) situações distintas que podem ocorrer quando um corpo rígido se encontra em contacto com uma superfície horizontal, rugosa e seca, conforme se esquematiza na Fig. 2:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P3/MAI_Ch03_P3ii.jpg"
width="900">

Para estes cenários poder-se-à analisar a reação como a resultante $\vec{R}$:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P3/MAI_Ch03_P3iii.jpg"
width="900">

Da análise da Fig. 3(c) pode-se obter a relação entre o ângulo de atrito estático $\phi_e$ com o coeficiente de atrito estático $\mu_e$, i.e.:

\begin{equation*}
\tan(\phi_e) = \frac{F_m}{N} = \frac{\mu_e N}{N} = \mu_e
\end{equation*}

De forma análoga, da Fig. 3(d), pode-se obter a relação entre o ângulo de atrito cinético $\phi_c$ com o coeficiente de atrito cinético
$\mu_c$:

\begin{equation*}
\tan(\phi_c) = \frac{F_c}{N} = \frac{\mu_c N}{N} = \mu_c
\end{equation*}
