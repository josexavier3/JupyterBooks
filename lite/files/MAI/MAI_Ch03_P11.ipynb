#### 1.2 Determine a intensidade e a direção da força que o cabo exerce sobre o centro da roldana D.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P11/MAI_Ch03_P11b.png"
width="400">


```python
import numpy as np
from numpy import cos, sin
import sympy as sy
from sympy.solvers import solve
from sympy.physics.mechanics import ReferenceFrame, dot

Tc = 111.0 # unit: N
t = sy.symbols('t')
ang = np.deg2rad(30)

#######################
SFx = -t + 2*t*cos(ang)
#######################
print(SFx)
tx = float(SFx.subs(t,Tc))
print(f'x: Fx={SFx} | {tx:.1f} [N]')
####################
SFy = -2*t*sin(ang)
####################
ty = float(SFy.subs(t,Tc))
print(f'y: Fy={SFy} | {ty:.1f} [N]')

S = ReferenceFrame('S') # ||x|| = ||y|| = ||z|| = 1
vT = tx*S.x + ty*S.y
print(vT)
Mod = np.sqrt(float(dot(vT,vT)))
print(f'Mod {Mod:.1f} [N]')
Ori = np.rad2deg(np.arctan(ty/tx))
print(f'Ori {Ori:.1f} [deg]')
```

#### 1.4 Calcule a força de atrito exercida na superfície do bloco que está em contacto com o plano inclinado.

```python
m1 = 30 # Unit: Kg
fa = sy.symbols('fa')
###############################
SFt = m1*g*sin(ang) - 2*Tc + fa
###############################
sol = solve(SFt,fa)
FA = sol[0]
print(f'SFt = 0 : {SFt}')
print(f'  FA = {FA:.1f} N')
```

Confirmar que $F_a < F_{a}^{max}$

```python

me = 0.5
n = sy.symbols('n')
###############################
SFn = n - m1*g*cos(ang)
###############################
sol = solve(SFn,n)
N = sol[0]
print(f'SFn = 0 : {SFn}')
print(f'  N = {N:.1f} N')
famax = me*N
print(f'famax = {famax:.1f} N')
```

#### 1.5 Determine o valor mínimo da massa do bloco m1min, i.e., quando existe iminência de escorregamento do mesmo.

Na iminência do escorregamento $F_a = F_{a}^{max}$

\begin{equation*}
F_a \equiv F_{a}^{max} = \mu N = \mu m_1 g \cos\theta
\quad\wedge\quad
\sum F_t = 0
\end{equation*}


```python
m_min = sy.symbols('m_min')
###############################
famax_mmin = me*(m_min*g*cos(ang))
print(f'famax_mmin = 0 : {famax_mmin}')
SFt_mmin = m_min*g*sin(ang) - 2*Tc + famax_mmin
print(f'SFt_mmin = 0 : {SFt_mmin}')
###############################
sol = solve(SFt_mmin,m_min)
Mm = sol[0]

print(f'  Mm = {Mm:.1f} Kg')
```
