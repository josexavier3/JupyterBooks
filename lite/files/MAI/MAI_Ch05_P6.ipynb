### Resolução

#### a) Diga o que entende por momento estático de uma superfície em relação a um eixo. ([_enunciatum_](#P6))

O momento estático ou momento de primeira ordem de uma superfície em relação a um eixo é uma medida da distribuição de massa ou área dessa superfície em relação a um eixo de referência. O momento estático é calculado multiplicando a distância de cada elemento infinitesimal da superfície até o eixo de referência pelo valor da área elementar nesse ponto e, em seguida, somando ou integrando esses produtos para toda a região: 

\begin{equation*}
Q_x = \int y\,\mathrm{d}A
\quad\wedge\quad
Q_y = \int x\,\mathrm{d}A
\end{equation*}

em que $y$ e $x$ representa a distância do elemento de área infinitesimal $\mathrm{d}A$ ao eixo em relação ao qual a expressão do momento estático é avaliado.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P6/MAI_Ch05_P6a.jpg"
style="max-width: 100%;"/>

O momento estático é uma quantidade importante na mecânica, usado para analisar sistemas em equilíbrio ou o comportamento de objetos em rotação.


#### b) Determine, utilizando as tabelas dadas, o momento estático da superfície representada em relação ao eixo $y$. ([_enunciatum_](#P6))

A figura enunciada pode ser decomposta em geometrias simples atendendo a áreas comuns (Anexo: Formulário). Neste caso, esta pode ser obtida pela adição de áreas correspondentes a um triângulo e um parto de círculo:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P6/MAI_Ch05_P6b.jpg"
style="max-width: 100%;"/>

O momento estático da superfície em relação ao eixo $y$ pode ser obtido a partir do conhecimento das coordenadas dos centros de gravidade de cada subárea que o constituem, de acordo com a seguinte expressão:

\begin{equation*}
Q_y = \overline{X} A = \sum_i \overline{x}_i A_i
\end{equation*}

Por consulta direta do formulário, resulta:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P6/MAI_Ch05_P6c.jpg"
width="350""/>

<div class="foo">

| Sub-áreas|$A_i$, m $^2$|$\overline{x}_i$, m|$\overline{x}_iA_i$, m $^3$|
|:-------:|:----------:|:-----------------:|:------------------------:|
| (1) [triângulo] | $\displaystyle\frac{2a\cdot a}{2} = a^2$ | $\displaystyle\frac{a}{3}$ | $\displaystyle\frac{a^3}{3}$ |
| (2) [4-círculo] | $\displaystyle\frac{2a^2}{3}$ | $\displaystyle\frac{3a}{8}$ | $\displaystyle\frac{a^3}{4}$ |
| $\sum : (1) + (2)$ | $\displaystyle\frac{5a^2}{3}$ | - | $\displaystyle\frac{7a^3}{12}$ |

</div>

Ou seja,

\begin{equation*}
\begin{aligned}
A &= \sum_{i=1}^2 A_i = a^2 + \frac{2a^2}{3} = \frac{5a^2}{3}~\,
(\mathrm{m}^2)\\
Q_y &= \sum_{i=1}^2 \overline{x}_i A_i =
\frac{a^3}{3} + \frac{a^3}{4} =
\frac{7a^3}{12}~\,(\mathrm{m}^3)
\end{aligned}
\end{equation*}

### c) Comprove o resultado da alínea anterior efectuando os cálculos por integração. ([_enunciatum_](#P6))

Para o desenvolvimento integral considere-se um elemento de área infinitesimal $\mathrm{d}A$ como o representado na figura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P6/MAI_Ch05_P6d.jpg"
width="350">

Por definição de momento estático (ou de primeira ordem de uma superfície) em relação ao eixo $y$:

\begin{equation*}
Q_y = \int x\,\mathrm{d}A = \int_{x_i}^{x_f}\int_{y_i(x)}^{y_f(x)} x\,
\mathrm{d}x\mathrm{d}y
\end{equation*}

Neste caso torna-se necessário definir os limites de integração da área - $x_i, x_f, y_i(x), y_f(x)$ - o que conduz ao seguinte desenvolvimento:

\begin{equation*}
\begin{aligned}
Q_y &= \int_{0}^{a}\int_{\displaystyle\scriptsize\frac{x^2}{a}-a}^{2(a-x)}x\,
\mathrm{d}y\mathrm{d}x =
\int_{0}^{a}\left(\int_{\displaystyle\scriptsize\frac{x^2}{a}-a}^{2(a-x)}x\,
\mathrm{d}y\right)\mathrm{d}x \\
&=\int_{0}^{a} \left[xy\right]_{\displaystyle\scriptsize\frac{x^2}{a}-a}^{2
(a-x)}\mathrm{d}x = \int_{0}^{a} x[2
(a-x)-\frac{x^2}{a}+a] \mathrm{d}x \\
&= \int_{0}^{a}\left(2ax-2x^2-\frac{x^3}{a} + xa\right) \mathrm{d}x\\
&= \left[ax^2-\frac{2x^3}{3}-\frac{x^4}{4a} + \frac{x^2a}{2}\right]_{0}^{a} =
\frac{7a^3}{12} \quad(\mathrm{m}^3)
\end{aligned}
\end{equation*}

```python
import sympy as sy

a, x, y = sy.symbols('a x y')

li = x**2/a-a
lf = 2*(a-x)
f = x
Qy = sy.integrate(f, (y,li,lf), (x,0,a))
print('Qy =',Qy,' m3')
```

Como complemento, a área da superfície pode ainda ser calcula por:

\begin{equation*}
\begin{aligned}
A &= \int \mathrm{d}A =
\int_{0}^{a}\int_{\displaystyle\scriptsize\frac{x^2}{a}-a}^{2(a-x)}\,
\mathrm{d}y\mathrm{d}x =
\int_{0}^{a}\left(\int_{\displaystyle\scriptsize\frac{x^2}{a}-a}^{2(a-x)}\,
\mathrm{d}y\right)\mathrm{d}x \\
&=\int_{0}^{a} \left[y\right]_{\displaystyle\scriptsize\frac{x^2}{a}-a}^{2
(a-x)}\mathrm{d}x = \int_{0}^{a} [3a-2x-\frac{x^2}{a}] \mathrm{d}x \\
&= \left[3ax-x^2-\frac{x^3}{3a}\right]_{0}^{a} =
\frac{5a^3}{3} \quad(\mathrm{m}^2)
\end{aligned}
\end{equation*}

```python
f = 1
A = sy.integrate(f, (y,li,lf), (x,0,a))
print('A =',A,' m')
```

### d) Enuncie os teoremas de Pappus-Guldinus. ([_enunciatum_](#P6))

**Teorema I**

A área de uma superfície de revolução é igual ao produto do comprimento da curva geratriz pelo caminho percorrido pelo centroide da curva durante o movimento de
rotação que gera a superfície:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P6/MAI_Ch05_P6e.jpg"
style="max-width: 100%;"/>

\begin{equation*}
A = \int \mathrm{d}A = \int 2\pi\,x\,\mathrm{d}L
= 2\pi \int x\,\mathrm{d}L = 2\pi\,\overline{X}\,L
\end{equation*}

**Teorema II**

O volume de um corpo de revolução é igual ao produto da área da superfície geratriz pelo caminho percorrido pelo centroide dessa superfície durante o
movimento de rotação que gera o corpo:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P6/MAI_Ch05_P6f.jpg"
style="max-width: 100%;"/>

\begin{equation*}
V = \int \mathrm{d}V = \int 2\pi\,x\,\mathrm{d}A
= 2\pi\underbrace{\int y\,\mathrm{d}A}_{Q_x}
= 2\pi\,\overline{X}\,A
\end{equation*}

### e) Considerando agora $a =$ 200 mm, determine o volume do sólido de revolução obtido por rotação da superfície em torno do eixo $y$. ([_enunciatum_](#P6))

Recorrendo ao 2º Teorema de Pappus-Guldinus ($V=2\pi\,\overline{X}\,A$), é preciso calcular a área ($A$) e a coordenada $x$ ($\overline{X}$) do centroide da figura:

\begin{equation*}
A = \frac{5a^2}{3} = \frac{5(0.2)^2}{3}
= 0.0667 \,(\textrm{m}^2)
\end{equation*}

```python
import numpy as np

Ae = A.subs(a,0.2)
print('A = ','{:.4f}'.format(Ae),' m2')
```

\begin{equation*}
\overline{X} = \frac{Q_y}{A},
\quad\mathrm{com}\quad
Q_y = \frac{7a^3}{12},~
A = \frac{5a^2}{3}
\end{equation*}

de onde resulta,

\begin{equation*}
\overline{X} = \frac{7a^3/12}{5a^2/3} = \frac{7a}{20}
= 0.07 \,(\textrm{m})
\end{equation*}

```python
Qye = Qy.subs(a,0.2)
print('Qy = ','{:.5f}'.format(Qye),' m3')
X = Qy/A
print('X  = ',X,' m')
Xe = X.subs(a,0.2)
print('X  = ','{:.4f}'.format(Xe),' m')
```

De onde o volume vem finalmente,

\begin{equation*}
V = 2\pi \underbrace{\overline{X}~\cdot~A}_{Q_y = 7a^2/12}
= 2\pi \left(\frac{7a}{20}\right)
\left(\frac{5a^2}{3}\right)
= \frac{7\pi a^3}{6} = 0.0293 \,(\textrm{m}^3)
\end{equation*}

```python
V = 2*np.pi*Xe*Ae
print('V = ','{:.4f}'.format(V),' m3')
```
