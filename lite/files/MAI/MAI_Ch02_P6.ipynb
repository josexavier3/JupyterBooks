#### b) Sabendo que $\alpha = 30^\circ$, determine a força de tração instalada em cada cabo.

Analisando o diagrama de corpo livre da partícula e assumindo equilíbrio estático têm-se:

\begin{equation*}
\sum\vec{F} = \vec{0}
\end{equation*}

Num dado sistema de coordenadas ortonormado, esta equação vetorial pode ser expressa pelas duas equações escalares seguintes, segundo as direções horizontal e vertical, respetivamente:

\begin{equation*}
\begin{cases}
\sum F_x &=& 0 \\
\sum F_y &=& 0
\end{cases}
\Leftrightarrow
\begin{cases}
T_\textrm{CB}\cos\alpha - T_\textrm{CA}\sin(5^\circ\frac{\pi}{180^\circ}) &=& 0
\\
-T_\textrm{CB}\sin\alpha + T_\textrm{CA}\cos(5^\circ\frac{\pi}{180^\circ}) - P
&=& 0
\end{cases}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

P = 5340 # unit: N
alpha = np.deg2rad(30) # unit: radianos
angAC = np.deg2rad(5) # unit: radianos

tca, tcb = sy.symbols('tca tcb')
EQ1 = tcb*np.cos(alpha)  - tca*np.sin(angAC)
EQ2 = -tcb*np.sin(alpha)  + tca*np.cos(angAC) - P
sol = solve({EQ1,EQ2},{tca, tcb})

TCA = float(sol[tca]); print('TCA = ','{:.1f}'.format(TCA),' N')
TCB = float(sol[tcb]); print('TCB = ','{:.1f}'.format(TCB),' N')
```

Numa abordagem matricial têm-se,

\begin{equation*}
\begin{bmatrix}
\cos\alpha & -\sin(5^\circ\frac{\pi}{180^\circ}) \\
-\sin\alpha & \cos(5^\circ\frac{\pi}{180^\circ})
\end{bmatrix}
\begin{bmatrix}
T_\textrm{CB} \\
T_\textrm{CA}
\end{bmatrix} =
\begin{bmatrix}
0 \\
P
\end{bmatrix}
\end{equation*}

```python
matA = np.array([[np.cos(alpha), - np.sin(angAC)],
                [-np.sin(alpha), np.cos(angAC)]])
print(f'A = {matA}')
matP = np.array([[0],[P]])
print(f'b = {matP}')
matT = np.linalg.inv(matA).dot(matP)
print(f'x = A^-1.b = {matT}')
```

Este resultado poderia ser obtido recorrendo à lei dos senos:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P6/MAI_Ch02_P6b.jpg"
style="width: 25%; max-width: 100%;"/>

\begin{equation*}
\begin{aligned}
&\frac{P}{\sin\left( 55^\circ\frac{\pi}{180^\circ} \right)} =
\frac{T_\textrm{CA}}{\sin\left( 120^\circ\frac{\pi}{180^\circ} \right)} =
\frac{T_\textrm{CB}}{\sin\left( 5^\circ\frac{\pi}{180^\circ} \right)}
\end{aligned}
\end{equation*}

```python
TCA = P*np.sin(np.deg2rad(120))/np.sin(np.deg2rad(55))
print('TCA = ', '{:.1f}'.format(TCA),' N')

TCB = P*np.sin(np.deg2rad(5))/np.sin(np.deg2rad(55))
print('TCB = ', '{:.1f}'.format(TCB),' N')
```

#### c) Sabendo que o ângulo que o troço AD do cabo faz com a horizontal é $\beta = 45^\circ$, determine a intensidade e direção da força exercida pelo cabo DAC sobre a roldana A.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch02/P6/MAI_Ch02_P6c.jpg"
style="width: 25%; max-width: 100%;"/>


Cálculo da resultante das duas forças ($F_x, F_y$) que atuam na roldana A
exercidas pelo
cabo:

\begin{equation*}
\begin{cases}
F_x = T_\textrm{AC}\sin(5^\circ\frac{\pi}{180^\circ}) - T_\textrm{AD}\cos
(45^\circ\frac{\pi}{180^\circ})
\\
F_y = -T_\textrm{AC}\cos(5^\circ\frac{\pi}{180^\circ}) - T_\textrm{AD}\sin
(45^\circ\frac{\pi}{180^\circ})
\end{cases}
\end{equation*}

De notar no entanto que a condição de roldana sem massa (desprezando o momento
de inércia) e sem atrito verifica-se:

\begin{equation*}
T_\textrm{AC} = T_\textrm{AD} = T
\end{equation*}

De onde resulta,

\begin{equation*}
\begin{cases}
F_x - T\left[\sin(5^\circ\frac{\pi}{180^\circ}) - \cos
(45^\circ\frac{\pi}{180^\circ})\right] = 0
\\
F_y + T\left[\cos(5^\circ\frac{\pi}{180^\circ}) + \sin
(45^\circ\frac{\pi}{180^\circ})\right] = 0
\end{cases}
\end{equation*}

```python
SumFx, SumFy = sy.symbols('SumFx SumFy')
EQ1 = SumFx - TCA*(np.sin(np.deg2rad(5)) - np.cos(np.deg2rad(45)))
EQ2 = SumFy + TCA*(np.cos(np.deg2rad(5)) + np.sin(np.deg2rad(45)))
sol = solve({EQ1,EQ2},{SumFx, SumFy})
Fx = float(sol[SumFx])
Fy = float(sol[SumFy])
print('Fx = ', '{:.2f}'.format(Fx),' N')
print('Fy = ', '{:.2f}'.format(Fy),' N')
```

De notar que a força $\vec{F}$ representada na figura é de igual em amplitude mas de sentido oposto (terceira lei de newton da ação-reação), à força exercida pela estrutura do camião na roldana A ($\sum\vec{F}=\vec{0}$).

**Vetor força resultante**

Tendo as componentes do vetor força no plano atuante na roldana A pela ação do cabo ($F_x,F_y$), é possível obter o módulo do vetor força resultante:

\begin{equation*}
||\vec{F}|| = F = \sqrt{\vec{F}\cdot\vec{F}} = \sqrt{F_x^2 + F_y^2}
\end{equation*}

```python
F = np.sqrt(Fx**2 + Fy**2)
print('|| F || = ', '{:.2f}'.format(F),' N')
```

bem como a orientação (ângulo em relação ao eixo vertical, por exemplo) do vetor resultante:

- Fõrmula de cálculo 1 (via trigonometria):

\begin{equation*}
\gamma = \arctan\left(\frac{F_x}{F_y}\right)
\end{equation*}

```python
print('expressao arctan (trigonometria): ----------------')
gam = np.arctan(Fx/Fy)
print('gam = ','{:.1f}'.format(np.rad2deg(gam)),' [graus]')
```

- Fórmula de cálculo 2 (via produto escalar):

\begin{equation*}
\gamma =
\arccos\left( \frac{\vec{F}\cdot(-\vec\jmath)}{||\vec{F}||} \right)
\end{equation*}

```python
print('expressao arccos (produto escalar): ----------------')
gam2 = np.arccos(-Fy/F)
print('gam = ','{:.1f}'.format(np.rad2deg(gam2)),' [graus]')
```
