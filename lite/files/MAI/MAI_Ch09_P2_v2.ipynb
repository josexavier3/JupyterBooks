Escolhe-se um sistema de coordenada $S(E,\vec{i},\vec{j})$ cuja origem seja fixa.

Represente-se o DCL do mecanismo (real), indicando sobre eles todas as forças atuantes.

A seguir, impõe-se uma perturbação na forma de uma rotação (virtual) $\delta\theta$ em E.

De notar que a existência da mola ligado ao sistema no ponto C, introduzirá uma força que pode ser positiva sobre o mecanismo se $+\delta\theta$ (mola em compressão) ou negativa se $-\delta\theta$ (mola em tração).

Desenha-se a configuração deformada virtual do sistema (conforme ilustrado na figura acima). Confrontando o sistema real (equilibrado e compatível) com a configuração virtual (compatível) identificam-se as forças e os momentos no sistema real que produzirão um trabalho virtual: $P$, $F_m$ e $M$.

Em E, as reações segundo $x$ e $y$ não realizam qualquer trabalho pois não existem quaisquer deslocamentos associados e em A o deslocamento virtual que lhe está associado é perpendicular à direção de $A_x$ pelo que, também não realiza trabalho (ou é nulo).

**Princípio dos trabalhos virtuais**

Antes de aplicar o PTV, define-se primeiro o deslocamento virtual em D ($\delta_{x_D}$) e em C ($\delta_{x_C}$):

\begin{equation}
\begin{cases}
x_D = 3L\cos\theta \\
x_C = L\cos\theta
\end{cases}
\end{equation}

Por derivação em relação a $\theta$:

\begin{equation}
\begin{cases}
\delta x_D = -3L\sin\theta\delta\theta \\
\delta x_C = -L\sin\theta\delta\theta
\end{cases}
\end{equation}

Aplicando o princípio dos trabalhos virtuais têm-se:

\begin{equation}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
\delta U_M + \delta U_P + \delta U_{F_M} = 0\\
&~\Leftrightarrow~
\vec{M} \cdot \delta\vec{\theta}
+ \vec{P} \cdot \vec\delta_D
+ \vec{F}_m \cdot \vec\delta_C = 0
\\
&~\Leftrightarrow~
M \cdot \delta\theta
+ P \cdot \delta_{x_D}
+ F_m \cdot \delta_{x_C} = 0
\\
&~\Leftrightarrow~
M - P(3L\sin\theta)
- F_m(L\sin\theta) = 0
\quad\wedge\quad \delta\theta \ne 0
\end{aligned}
\end{equation}

Para uma posição qualquer, e considerando que quando $\theta = 60^\circ$ a mola está indeformada (de acordo com o enunciado), a força na mola é defina ainda através da seguinte expressão:

\begin{equation}
F_m = k [x_C - L\cos(60^\circ)] \quad\wedge\quad
x_C = L\cos\theta
\end{equation}

Simplificando,

\begin{equation}
F_m = k [L\cos\theta - L\cos(60^\circ)]
= kL [\cos\theta - 1/2]
\end{equation}

```python
import numpy as np
import matplotlib.pylab as plt

# considerem-se valor arbitrários para as variáveis
L = 1
k = 1
aEQdeg = 60
aEQrad = np.deg2rad(aEQdeg)

def fmola(a): return k*L*(np.cos(a) - np.cos(aEQrad))

plt.rcParams['axes.facecolor'] = (.15, .15, .15)
plt.rcParams['figure.facecolor'] = (.15, .15, .15)
plt.rcParams['font.family'] = 'monospace'
plt.rcParams['font.size'] = 12
# plt.rcParams['text.usetex'] = True
params = {"ytick.color" : (.8, .8, .8),
          "xtick.color" : (.8, .8, .8),
          "grid.color" : (.2, .2, .2),
          "text.color" : (.7, .7, .7),
          "axes.labelcolor" : (.8, .8, .8),
          "axes.edgecolor" : (.15, .15, .15)}
plt.rcParams.update(params)

vart = np.arange(0,91,5)
plt.figure(figsize=(9,6))
ax = plt.gca()
xx1, yy1 = vart, fmola(np.deg2rad(vart))
plt.plot(xx1, yy1,'-',linewidth=4,color='brown',alpha=.7)
xt1 = vart[int(np.where(vart == 15)[0])]
yt1 = fmola(np.deg2rad(xt1))
plt.text(xt1,yt1*1.1,f'mola em tração',fontsize=20)
xt2 = vart[int(np.where(vart == 80)[0])]
yt2 = fmola(np.deg2rad(xt2))
plt.text(xt2*.55,yt2*1.5,f'mola em compressão',fontsize=20)
ax.axhline(y=0, color=(0.2,0.2,0.2))
ax.axvline(x=0, color=(0.9,0.9,0.9))
plt.plot(aEQdeg, fmola(aEQrad),'o',markeredgecolor='r',
                      markerfacecolor='g',markersize=22,alpha=0.6)
plt.xlim(0,90)
plt.ylim(-L/2*1.2,L/2*1.2)
plt.xlabel(r"$\theta$ (graus)", size=18)
plt.ylabel(r"$F_m$ (N)", size=18)
plt.show()
```

De notar que esta relação garante a condição de que quando $\theta =
60^\circ$, a força exercida pela mola é nula.

Pelo que o PTV:

\begin{equation}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
M - 3PL\sin\theta
- k L^2 [\cos\theta - 1/2]\sin\theta = 0\\
&~\Leftrightarrow~
M = 3PL\sin\theta
+ k L^2 [\cos\theta - 1/2]\sin\theta
\end{aligned}
\end{equation}

```python
P = 1

def mom1(a): return 3*P*L*np.sin(a)
def mom2(a): return 3*P*L*np.sin(a) + fmola(a)*L*np.sin(a)

vara = np.arange(0,91,5)
plt.figure(figsize=(9,6))
ax = plt.gca()
xx1, yy1 = vara, mom1(np.deg2rad(vara))
plt.plot(xx1, yy1,'--',linewidth=2,color='brown',alpha=.5,label='P2')
xx2, yy2 = vara, mom2(np.deg2rad(vara))
plt.plot(xx2, yy2,'-',linewidth=4,color='green',alpha=.5,label='P2v2')
ax.axhline(y=0, color=(0.2,0.2,0.2))
ax.axvline(x=0, color=(0.9,0.9,0.9))
ax.legend(('P2','P2v2'), loc='upper center', frameon=False, ncol=2)
plt.xlim(0,90)
plt.xlabel(r"$\theta$ (graus)", size=18)
plt.ylabel(r"$M$ (N.m)", size=18)
plt.show()
```
