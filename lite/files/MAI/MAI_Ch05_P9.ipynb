## CENTROIDES E CENTROS DE GRAVIDADE

### Problema 9

Considere a superfície plana cuja fronteira é definida por duas retas horizontais $y=a$ e $y=h$ e por duas retas que passam na origem e com inclinação $\alpha=60^\circ$ com o eixo horizontal.

a)	Determine, efetuando os cálculos por integração, o momento estático da superfície representada em relação ao eixo $x$.

b)	Determine, justificando, as coordenadas do centroide da superfície no referencial $x,y$ indicado.

c)	Determine, justificando, o volume obtido pela rotação completa da superfície em torno do eixo $x$.

### Resolução

#### a)	Determine, efetuando os cálculos por integração, o momento estático da superfície representada em relação ao eixo $x$.


**Equações das rectas que definem a região de interesse**

\begin{equation*}
y_1(x) = - \tan(\theta) x
\quad\wedge\quad
\theta=60^\circ
\quad\wedge\quad
\tan(60^\circ)= \sqrt{3}
\end{equation*}

pelo que resulta,

\begin{equation*}
x = - \frac{\sqrt{3}}{3} y
\end{equation*}

\begin{equation*}
y_2(x) = \tan(\theta) x
\quad\therefore\quad
x = \frac{\sqrt{3}}{3} y
\end{equation*}

**Área da região de interesse**

\begin{equation*}
A = \int \mathrm{d}A = \int_a^h \int_{-\sqrt{3}/3y}^{\sqrt{3}/3y} \mathrm{d}x\mathrm{d}y
= \int_a^h \left[ x \right]_{-\sqrt{3}/3x}^{\sqrt{3}/3x} \mathrm{d}y=
\int_a^h
2\frac{\sqrt{3}}{3}y
\mathrm{d}y=
\frac{2\sqrt{3}}{3}
\left[\frac{y²}{2}\right]_a^h
= \frac{\sqrt{3}}{3}(h²-a²)
\end{equation*}

**Primeiro momento de área**

\begin{equation*}
Q_x = \int\int y  \mathrm{d}A = \int_a^h \int_{-\sqrt{3}/3y}^{\sqrt{3}/3y} y\mathrm{d}x\mathrm{d}y
= \int_a^h y
\left(2\frac{\sqrt{3}}{3}y\right)
\mathrm{d}y=
\frac{2\sqrt{3}}{3}
\left[\frac{y³}{3}\right]_a^h
= \frac{2\sqrt{3}}{9}(h²-a²)
\end{equation*}


#### b)	Determine, justificando, as coordenadas do centroide da superfície no referencial $x,y$ indicado.

\begin{equation*}
\overline{y} = \frac{Q_x}{A} = \frac{\frac{2\sqrt{3}}{9}(h²-a²)}{\frac{\sqrt{3}}{3}(h²-a²)}
= \frac{2}{3}\left(\frac{h^2-a³}{h²-a²}\right)
\end{equation*}

\begin{equation*}
\overline{x} = 0
\quad\textrm{(existe simetria em relação ao eixo $y$)}
\end{equation*}



#### c)	Determine, justificando, o volume obtido pela rotação completa da superfície em torno do eixo $x$.

\begin{equation*}
V = \int \textrm{d}V = \int 2 \pi y \textrm{d}A = 2 \pi \int y \textrm{d}A = 2\pi Q_x = 2\pi \frac{2\sqrt{3}}{9}
\left(h³-a³\right)
= 4\pi\frac{\sqrt{3}}{9}
\left(h³-a³\right)
\end{equation*}
