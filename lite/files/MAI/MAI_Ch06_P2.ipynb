#### a)  Determine as coordenadas do centroide da superfície no referencial $S (O,x,y)$ indicado. ([_enunciatum_](#P4))

Uma vez que o eixo dos $y$ é um eixo de simetria, conclui-se diretamente que $Q_y = 0~\therefore~\overline{x} = 0$ mm.

Para definir as coordenadas do centroide da figura, comece-se por decompor a região em duas outras mais simples conforme se ilustra a figura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P2/MAI_Ch06_P2a.jpg"
width="900">

Para figuras elementares as coordenadas do centroide e área são obtidos diretamente por consulta a tabelas (Anexo A):

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P2/MAI_Ch06_P2b.jpg"
width="900">

Por conseguinte, as coordenada do centroide da superfície pode ser calculada pela seguinte expressão:

\begin{equation*}
\overline{y} = \displaystyle\frac{\sum_i y_i A_i}{\sum_i A_i}
\quad\wedge\quad
\overline{x} = 0
\end{equation*}

**Semi-elipse**

- Centroides \& Área da semielipse (tabela)

    \begin{equation*}
    \overline{x}_1 = 0
    \quad\wedge\quad
    \overline{y}_1 = \displaystyle\frac{4b}{3 \pi}
    \quad\wedge\quad
    A_1 = \displaystyle\frac{\pi ab}{2}
    \end{equation*}

```python
import numpy as np

b = 36 #  units: mmm
a = 54

y1 = 4*b/3/np.pi
print(f'y1 = {y1:.1f} mm')
A1 = np.pi*a*b/2
print(f'A1 = {A1:.1f} mm2')
```

**Semi-círculo**

- Centroides \& Área da semielipse (tabela)

    \begin{equation*}
    \overline{x}_2 = 0
    \quad\wedge\quad
    \overline{y}_2 = \displaystyle\frac{4r}{3 \pi}
    \quad\wedge\quad
    A_2 = \displaystyle\frac{\pi r^2}{2}
    \end{equation*}

```python
r = 18
y2 = 4*r/3/np.pi
print(f'y2 = {y2:.1f} mm')
A2 = np.pi*r**2/2
print(f'A2 = {A2:.1f} mm2')
```

Em síntese,

| Áreas:&nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|$A_i$, mm $^2$&nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|$\overline{y}_i$, mm&nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|$\overline{y}_iA_i$, mm $^3$&nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|
|:-------:|:----------:|:-----------------:|:------------------------:|
|`(1)`|$\displaystyle\frac{\pi ab}{2}$|$\displaystyle\frac{4b}{3\pi}$|$\displaystyle\frac{2ab^2}{3}$|
|`(2)`|$\displaystyle\frac{\pi r^2}{2}$|$\displaystyle\frac{4r}{3 \pi}$|$\displaystyle\frac{2r^3}{3}$|
|$\sum: (2) - (1)$|$\displaystyle\frac{\pi}{2}(ab - r^2)$|-|$\displaystyle\frac{2}{3}(ab^2- r^3)$|

Pelo que resulta,

\begin{equation*}
\begin{aligned}
A & = A_1 - A_2 = \frac{\pi}{2}(ab - r^2)\\
Q_x & = \overline{y}_1A_1 - \overline{y}_2A_2
=\frac{2ab^2}{3} - \frac{2r^3}{3}
=\frac{2}{3}(ab^2- r^3)\\
\overline{y} &= \frac{Q_x}{A} =
\frac{4}{3\pi}\left(\frac{ab^2- r^3}{ab - r^2}\right)
= 16.8~\textrm{mm}\\
\end{aligned}
\end{equation*}

```python
print('Cal 1: ------------------')
At = A1 - A2
print(f'A = {At:.1f} mm2')
Qx = y1*A1 - y2*A2
print(f'Qx = {Qx:.1f} mm3')
ycoord  = Qx/At
print(f'ycoord = {ycoord:.2f} mm')
```

```python

print('\nCal 2: ------------------')

import sympy as sy
from sympy.solvers import solve
from sympy.physics.mechanics import ReferenceFrame, Vector, dot, cross

R, A, B = sy.symbols('R A B')

def semielipse(a, b):
    y = 4*b/3/np.pi
    A = np.pi*a*b/2
    return y, A

def semicirculo(r):
    y = 4*r/3/np.pi
    A = np.pi*r**2/2
    return y, A

ycirc1, Acirc1 = semielipse(A,B)
ycirc2, Acirc2 = semicirculo(R)

sA = Acirc1 - Acirc2
print(f'A = {sy.factor(sA)} = {sA.subs([(A,a),(B,b),(R,r)]):.1f} mm2')
sQy = ycirc1*Acirc1 - ycirc2*Acirc2
print(f'Qy = {sy.factor(sQy)} = {sQy.subs([(A,a),(B,b),(R,r)]):.1f} mm3')
sy = sQy/sA
print(f'y = {sy.factor(sy)} = {sy.subs([(A,a),(B,b),(R,r)]):.3f} mm')
```

#### b)  Determine o momento polar de inércia da superfície relativamente ao centroide. ([_enunciatum_](#P4))

Para calcular o momento de inércia de formas de superfície pode-se recorrer à tabela do Anexo B.

Recorrendo novamente à decomposição das figuras [1] e [2], proposta na alínea anterior:

_Momento polar da semi-elipse:_

O momento polar da semi-elipse relativamente ao Ponto O (origem do sistema de eixos cartesiano $S(0,\vec{\imath},\vec{\jmath})$) vem:

\begin{equation*}
\begin{aligned}
J_0^{(1)} &= \frac{1}{2}\left[\frac{1}{4}\pi a b (a^2 + b^2)\right]
\end{aligned}
\end{equation*}

Note-se que a tabela apresenta um formulãrio para a elipse completa. Assim sendo, é necessário dividir o resultado por dois para considerar efetivamente metade da elipse.

Note-se também, por conveniência, que o sistema de coordenadas escolhido $S(0,
\vec{\imath},
\vec{\jmath})$ é coincidente com o referencial usado nas tabelas.

```python
def J0semielipse(a,b):
    return 0.5*(1/4*np.pi*a*b*(a**2+b**2))

J01 = J0semielipse(A,B)
print(f'J01 (semielipse) = {J01} = {J01.subs([(A,a),(B,b)]):.3e} mm4')
```

_Momento polar da semi-círculo:_

Por sua vez, o momento polar de inércia do semi-círculo relativamente ao Ponto O é:

\begin{equation*}
\begin{aligned}
J_0^{(2)} &= \frac{1}{4}\pi r^4
\end{aligned}
\end{equation*}

Mais uma vez, salienta-se a coincidência do sistema de coordenadas escolhido para os cálculos e o usado na tabela.

```python
def J0semicirculo(r):
    return 1/4*np.pi*r**4

J02 = J0semicirculo(R)
print(f'J02 (semicirculo) = {J02} = {J02.subs([(R,r)]):.3e} mm4')
```

_Momento polar de inércia da figura:_

Assim, o momento polar de inércia da figura relativamente ao Ponto O vale

\begin{equation*}
J_0 = I_x + I_y = J_0^{(1)} - J_0^{(2)}
= \frac{\pi}{8} (a^3 b + a b^3 - 2 r^4)
\end{equation*}

```python
J0 = J01 - J02
print(f'J0 (J01 - J02) = {J0} = {J0.subs([(A,a),(B,b),(R,r)]):.3e} mm4')
```

_Teorema dos eixos paralelos:_

De notar que o momento polar de inércia calculado anteriormente é relativo ao sistema de coordenadas global $S(0,\vec{\imath},\vec{\jmath})$. Interessa-nos contudo expressar esta valor em relação a eixos de um sistema de coordenadas com origem no centroide da região original.

Aplicando agora o teorema dos eixos paralelos, o momento polar de inércia relativamente ao centroide da figura calcula-se de acordo com:

\begin{equation*}
J_0 = J_c + Ad^2
\end{equation*}

onde

- $A$ é a área total da figura;
- $J_C$ é o momento polar de inércia relativamente ao centroide da figura;
- $d$ é a distância entre os eixos que passam pelos Pontos O e C.

Desta forma, o momento polar de inércia da figura relativamente ao Ponto C (centroi da figura) vem:

\begin{equation*}
J_c = J_0 - Ad^2
= J_0 - A\overline{y}^2
= \frac{1}{8} a b (a^2 + b^2)\pi - \frac{\pi r^4}{4} -
\frac{8 (r^3 - ab^2)^2}{9\pi (a b - r^2)}
\end{equation*}

```python
J0val = J0.subs([(A,a),(B,b),(R,r)])
JC = J0val - At*ycoord**2
print(f'JC = {JC:.3e} mm4')

def jcanal(a,b,r):
    c1 = 1/8*a*b*(a**2 + b**2)*np.pi
    c2 = np.pi*r**4/4
    c3 = (8*(r**3 - a*b**2)**2)/(9*np.pi*(a*b-r**2))
    return c1 - c2 - c3

JCa = jcanal(a,b,r)
print(f'JC (analitico) = {JCa:.3e} mm4')
```
