### Resolução

#### a) As reações nos apoios. <a id="A"></a>([_enunciatum_](#PII))

Desenha-se primeiro o DCL da treliça

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch07/P2/MAI_Ch07_P2a.jpg"
width="700">

A treliça contem:

| N.º de nós, $n$ | N.º de barras, $m$ | N.º de reações, $r$ |
| ---------- | ------------- | -------------- |
| 11 | 19 | 3 |

- Da análise da estrutura isostática exterior conclui-se: $r = 3$.
- Da análise da estrutura isostática interior conclui-se: $m = 2n-3$.
- Da análise global da estrutura permite concluir que a treliça é
estaticamente determinada:

\begin{equation*}
g = 2n - (m+r) = 0
\end{equation*}

em que,

\begin{equation*}
\begin{aligned}
g &: \textrm{nº de graus de liberdade} \\
2n &: \textrm{nº máx. de equações independentes} \\
m+r &: \textrm{nº total de incógnitas}
\end{aligned}
\end{equation*}

```python
n, m, r = 11, 19, 3
print(f'2n = {2*n:.0f} | m  = {m:.0f} | r  = {r:.0f}')

def grauhiperstaticidade(nn,mm,rr):
    return 2*nn - (mm+rr)

g = grauhiperstaticidade(n,m,r)
print(f'g  = {g:.0f}')
```


**Estatia global**

Estatia global da treliça ($\alpha_g$) é calculada de acordo com:

\begin{equation*}
\alpha_g = \alpha_e + \alpha_i
\end{equation*}

em que $\alpha_e$ e $\alpha_i$ são a estatia exterior e estatia interior da treliça calculadas, respectivamente, por:

\begin{equation*}
\alpha_e = 3 × \textrm{( nº de apoios − 1)} − \textrm{nº de libertações exteriores}
\end{equation*}

- Apoios: O número de apoios refere-se ao total de apoios que a treliça tem
- Libertações Exteriores: Este termo refere-se ao número de graus de liberdade que não estão restringidos pelos apoios

\begin{equation*}
\alpha_i = 3 × \textrm{nº de malhas fechadas} − \textrm{nº de ligações interiores}
\end{equation*}

- Malha fechada numa treliça refere-se a uma região da estrutura que é completamente delimitada por barras, formando um circuito fechado. Malhas fechadas considera que cada malha adiciona três graus de liberdade internos.
- Ligações interiores referem-se às barras internas da treliça que conectam os nós entre si.

```python
def estatiaexterior(napoios, nlibertacoes):
    return 3 * ( napoios - 1 ) - nlibertacoes

def estatiainterior(nmalhas,nligainter):
    return  3 * nmalhas - nligainter

npoios_, nlibext_ = 2, 3
alphe = estatiaexterior(npoios_, nlibext_)
print(f'estatia exterior = 3({npoios_} - 1) - {nlibext_} =  {alphe}')
nmalh_, nlibint_ = 9, 1*2 + 2*2 + 3*7 
alphi = estatiainterior(nmalh_,nlibint_)
print(f'estatia interior = 3({nmalh_}) - {nlibint_} = {alphi}')
estatiaglobal = alphe + alphi
print((f'estatia global = {estatiaglobal}'))
```

**Equilíbrio da treliça inteira - determinação das reações**

As reações nos apoios podem ser calculadas recorrendo às equações de equilíbrio seguintes:

\begin{equation*}
\begin{cases}
\rightarrow(+)&:&\sum F_x = 0 \\
\uparrow(+)&:&\sum F_y = 0 \\
\circlearrowleft(+)&:&\sum M_A =0
\end{cases}~\Leftrightarrow~
\begin{cases}
R_{Ax} &=~0 \\
R_{Ky} + R_{Ay} - 2F &=~0 \\
-aF - 2aF + 5aR_{Ky} &=~0
\end{cases}
\end{equation*}

Resultando,

\begin{equation*}
\begin{cases}
R_{Ax} = 0 \\
R_{Ay} = \frac{7}{5}F \\
R_{Hy} = \frac{3}{5}F
\end{cases}
\end{equation*}

```python
import sympy as sy
from sympy.solvers import solve

# unit: m, kN
LX, LY, F = 3.81, 4.57, 26.7

rax, ray, rky, f, a = sy.symbols('rax ray rky f a')

SFX = rax
print('sum Fx :', SFX)
SFY = rky + ray - 2*f
print('sum Fy :', SFY)
SMA = -a*f - 2*a*f + 5*a*rky
print('sum MA :', SMA)
sol = solve({SFX,SFY,SMA},{rax, ray, rky})

RAx = sol[rax]
NRAx = RAx.subs({(f,F),(a,LX)})
print('RAx    = ', RAx,' kN')
# print('N[RAx] = ', NRAx,' kN')
RAy = sol[ray]
NRAy = RAy.subs({(f,F),(a,LX)})
print(f'RAy    = {RAy} | N[RAy] = {NRAy:.2f} kN')
RKy = sol[rky]
NRKy = RKy.subs({(f,F),(a,LX)})
print(f'RKy    = {RKy} | N[RKy] = {NRKy:.2f} kN')
```

#### b) Os esforços nas barras CE, DE e DF utilizando o método das secções. <a id="B"></a>([_enunciatum_](#PII))

**Método das Secções**

O método dos nós é utilizado quando se pretende determinar as forças em todos os elementos de uma treliça. Se, no entanto, somente se desejar conhecer as forças em um ou alguns elementos, o método das secções é mais indicado.

Para determinar a força nos elementos, pode-se considerar uma secção através de três elementos da treliça  dividindo-a em duas partes separadas. De notar, no entanto, que esta secção não  pode cortar mais do que três elementos (3 incógnitas). Cada uma das duas partes  da treliça obtidas pode ser utilizada como um corpo rígido.

_Diagrama de corpo livre da sub-treliça_

Para se calcularem os esforços nas barras CE, DE e DF recorrendo ao método das secções, faz-se um corte mm’ conforme a figura abaixo, i.e. cortando as barras CE, DE e DF. Desenha-se o DCL da parte esquerda da treliça e procede-se ao seu equilíbrio.


<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch07/P2/MAI_Ch07_P2b.jpg"
width="800">

Em conformidade com a terceira lei de Newton que afirma que, as forças de ação e reação entre corpos em contacto têm a mesma intensidade, a mesma linha de acção, e sentidos opostos, forças com sentido para fora dos nós indicam tração das barras sobre esse nó, e que, inversamente, forças que entram no nó, indicam que a barra associadas está à compressão.

_Estudo do equilíbrio estático da sub-estrutura_

Aplique-se as equações de equilíbrio estático à sub-estrutura:

- Neste caso temos há interesse em analisar: $\sum F_x = 0$.
- Por outro lado, pode-se usar a expressão do equilíbrio dos momentos escolhendo os pontos D e E.

Este desenvolvimento conduz a ($R_{Ax} = 0$):

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum M_D = 0 \\
\sum M_E = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
F_{DF} + F_{CE} + F_{DE}\cos\alpha = 0 \\
bF_{CE} + \frac{a}{2}F
- \left(\frac{a}{2} + a\right)R_{Ay} = 0 \\
aF - 2aR_{Ay} - bF_{DF} = 0
\end{cases}
\end{equation*}

\begin{equation*}
~\Leftrightarrow~
\begin{cases}
F_{DE} = -\frac{F_{DF} + F_{CE}}{\cos\alpha} \\
F_{CE} = \frac{a}{b}\left
(\frac{3}{2}R_{Ay} - \frac{1}{2}F\right) \\
F_{DF} = \frac{a}{b}(F-2R_{Ay})
\end{cases}
\end{equation*}

com: $\alpha = \arctan[b/(a/2)] = \arctan(2L_Y/L_X)$.

Resultando,

\begin{equation*}
\begin{cases}
F_{DE} = 11.57 ~\textrm{kN (T)} \\
F_{CE} = 35.62 ~\textrm{kN (T)} \\
F_{DF} = -40.07 ~\textrm{kN (C)}
\end{cases}
\end{equation*}

```python
import numpy as np

alpha = np.arctan(2*LY/LX)
print(f'alpha = {alpha:.3f} radianos = {np.rad2deg(alpha):.1f} graus')

fdf, fce, fde, t, b = sy.symbols('fdf fce fde t b')

SFX = fdf + fce + fde*sy.cos(t)
print('sum Fx :', SFX)
SMD = b*fce + a/2*f - (a/2+a)*ray
print('sum MD :', SMD)
SME = a*f - 2*a*ray - b*fdf
print('sum DE :', SME)
sol = solve({SFX,SMD,SME},{fdf, fce, fde})

FDF = sol[fdf]
NFDF = FDF.subs({(t,alpha),(ray,NRAy),(f,F),(a,LX),(b,LY)})
print(f'FDF    = {FDF} = {NFDF:.2f} kN (C)')

FCE = sol[fce]
NFCE = FCE.subs({(t,alpha),(ray,NRAy),(f,F),(a,LX),(b,LY)})
print(f'FCE    = {FCE} = {NFCE:.2f} kN (T)')

FDE = sol[fde]
NFDE = FDE.subs({(t,alpha),(ray,NRAy),(f,F),(a,LX),(b,LY)})
print(f'FDE    = {FDE} = {NFDE:.2f} kN (T)')
```
