A figura representa duas vigas ABC e CD, ligadas entre si por uma articulação em
C. A viga ABC está encastrada em A e a viga CD tem um apoio móvel em D. Para os
carregamentos representados na figura, determine:

a)	As reações nos apoios A e D.

b)	As equações do esforço transverso e do momento fletor para o troço CD.

c)	Os diagramas dos esforços internos para as duas vigas.

#### Resumo

**Determinação de Esforços em vigas**

Metodologia:

- Determinar reações nos apoios;
- Determinar os esforços internos (esforço transverso e momento
flector);
- Desenhar os diagramas de esforço transverso e momento flector.

Programas de análise estrutural:

- [Ftool](https://www.ftool.com.br/Ftool/)
- [ANSYS Student](https://www.ansys.com/academic/free-student-products)
- [Robot Structural Analysis Professional](https://www.autodesk.com/education/free-software/featured)

#### Resolução

```python
import numpy as np
import sympy as sp

h: float = 2.
v: float = 3.
FB: float = 500.
MB: float = 300.
dist: float = 250.

ang = np.arctan(v/(2*h))
print(f'{ang = :.2f} = {np.rad2deg(ang):.2f}')
ax, ay, tc = sp.symbols("ax ay tc")
SFX = sp.Eq(ax - tc*np.cos(ang),0)
SFY = sp.Eq(ay - FB - dist*h + tc*np.sin(ang),0)
SMA = sp.Eq(-MB - dist*h*(h/2) - FB*h + tc*np.sin(ang)*(2*h),0)
sol = sp.solve([SFX,SFY,SMA],[ax,ay,tc])
print(sol)
AX = sol[ax]
AY = sol[ay]
TC = sol[tc]
print(f'{AX = :.2f} [N]')
print(f'{AY = :.2f} [N]')
print(f'{TC = :.2f} [N]')
TCx = TC*np.cos(ang)
TCy = TC*np.sin(ang)
print(f'{TCx = :.2f} [N]')
print(f'{TCy = :.2f} [N]')
```
