### Resolução

Neste problema analisar-se-á a aplicação do princípio dos trabalhos virtuais em relação a um momento aplicado. Neste sistema, a aplicação da força horizontal $P$ provoca um movimento lateral, permitido pelo movimento vertical do pino em A. Por sua vez, este movimento pode ser equilibrado pela aplicação de um momento $M$ aplicado em E. Este problema pode ser analisado considerando apenas 1 (um) grau de liberdade, neste caso o ângulo $\theta$, na medida em
que todas os restantes deslocamentos se podem exprimir à custa desta grandeza.

**Diagram de corpo livre da estrutura**

Represente-se o DCL do mecanismo (real), indicando sobre eles todas as forças atuantes:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P2/MAI_Ch09_P2a.jpg"
width="400">

Considere-se um sistema de coordenadas com origem no ponto fixo E, $S(E, \vec{i},\vec{j})$. Considere-se a estrutura real numa configuração de equilíbrio estático.

De notar que uma variação de rotação do braço EB em torno de E no sentido anti-horário ($+\delta\theta$) irá provocar um deslocamento horizontal em D negativo ($-x_D$) e, inversamente, a variação de $-\delta\theta$ (sentio horário) irá provar um deslocamento positivo do ponto  de aplicação da força $P$, ou seja, $+x_D$.

**Movimento virtual admissível, $\delta\theta$**

Considere-se agora um movimento virtual do mecanismo ou variação do sistema em relação à sua configuração inicial de equilíbrio. Se o deslocamento virtual considerado for consistente com as condições de fronteira ou restrições impostas pelos suportes e conexões, todas as reações e forças são eliminadas da análise pelo princípio dos trabalhos virtuais. Apenas o trabalho das forças externas diretamente aplicadas, bem como forças de atrito precisam tipicamente de ser consideradas.

**Identificação das forças externas que realizam trabalho virtual**

Assuma-se um deslocamento virtual consistente com um deslocamento horizontal do ponto D, onde é aplicada a carga $P$.

-A força de reação em A, $A_x$, faz um ângulo reto ($90^\circ$) com o deslocamento vertical desse ponto, pelo que o seu trabalho virtual será nulo ($\cos 90^\circ = 0$).

- O ponto E é fixo pelo que as reações nesse ponto não realização trabalho virtual.

- Considera-se desprezável o peso próprio do mecanismo face ao carregamento externo a que está submetido.

- As forças externas aplicadas à estrutura que realizam trabalho virtual são a força horizontal em D ($P$) e o momento fletor em E ($M$).

**Princípio dos trabalhos virtuais**

Aplicando o princípio dos trabalhos virtuais têm-se:

\begin{equation*}
\delta U = 0~\textrm{[unidades: J $\equiv$ Nm]}
\end{equation*}

\begin{equation*}
\delta U = \delta U_M + \delta U_P = 0
~\Leftrightarrow~
\vec{M}\cdot\delta\vec{\theta} + \vec{P}\cdot\vec{\delta}_D = 0
\end{equation*}

_Vetores força e momento_

\begin{equation*}
\vec{M} = M~\vec{k}
\quad\wedge\quad
\vec{P} = P~\vec{\imath}
\end{equation*}

_Deslocamento virtual em D_

A variação ângular $\delta\vec{\theta}$ tem direção segundo o eixo de rotação:

\begin{equation*}
\vec{\delta\theta} = \delta\theta~\vec{k}
\end{equation*}

Este problema tem apenas um grau de liberdade independente, uma vez que os deslocamentos virtuais do sistema podem ser relacionados entre si geometricamente:

\begin{equation*}
x_D = 3L\cos\theta
\end{equation*}

Uma variação da configuração de equilíbrio por uma pequena rotação $\delta\theta$ irá provocar uma variação da abscissa do ponto D ($\delta x_D$) que pode ser calculada pela derivada de $x_D$ em relação à $\theta$:

\begin{equation*}
\begin{aligned}
\delta x_D = -3~L\sin\theta\,\delta\theta
&\quad\therefore\quad
\vec{\delta}_D = \left(-3~L\sin\theta\,\delta\theta\right)\vec{\imath}
\end{aligned}
\end{equation*}

De notar que a atratividade do princípio dos trabalhos virtuais depende em
grande parte da existência de relações geométricas simples entre os vários
deslocamentos virtuais envolvidos na solução de um problema.

Desenvolvendo,

\begin{equation*}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
M\delta\theta
- P(3L\sin\theta\delta\theta) = 0\\
&~\Leftrightarrow~
M- 3PL\sin\theta = 0
\quad\wedge\quad \delta\theta \ne 0 \\
&~\Leftrightarrow~
M = 3LP\sin\theta
\end{aligned}
\end{equation*}

```python
P = 1

def mom(a): return 3*P*L*np.sin(a)

vara = np.arange(0,91,5)
plt.figure(figsize=(9,6))
ax = plt.gca()
xx1, yy1 = vara, mom(np.deg2rad(vara))
plt.plot(xx1, yy1,'-',linewidth=4,color='brown',alpha=.5)
plt.xlim(0,90)
plt.xlabel(r"$\theta$ (graus)", size=18)
plt.ylabel(r"$M$ (N.m)", size=18)
plt.show()
```
