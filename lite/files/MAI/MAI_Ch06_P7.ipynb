#### a) Determine o momento polar de inércia da superfície em relação à origem do referencial, $J_0$. <a id="A"></a> ([_enunciatum_](#P7))

Não sendo explicitado nenhum caminho de resolução, opte-se por analisar a figura original como resultado da decomposição de figuras simples, cujos
dados figuram convenientemente em tabelas de consulta em anexo (Anexo A).

Começa-se por considerar a decomposição da figura inicial conforme se propõe na figura abaixo:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P7/MAI_Ch06_P7a.jpg"
width="700">

Para determinar o momento polar da figura original em relação à origem do referencial $S(0,\vec{\imath},\vec{\jmath})$ ($J_0$), pode ser conveniente, pelo menos para algumas figuras geométricas ($i$), calcula-se em alternativa, os momentos de 2ª ordem relativamente aos eixos $x$ ($I_x^i$) e $y$ ($I_y^i$). O formulário para a resolução da questão é dado por:

\begin{equation*}
J_0 = J_0^\Diamond - J_0^\blacksquare - J_0^\square
\quad\wedge\quad J_0^\Diamond = I_x + I_y
\quad\wedge\quad J_0^\blacksquare,~J_0^\square~\textrm{(diretamente da tabela)}
\end{equation*}

**Losango (2 triângulos)**

Da consulta direta do formulário para os momentos de inércia da figura de um triângulo resulta:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P7/MAI_Ch06_P7b.jpg"
width="600">

As sub-área do triângulo para o cálculo do momento de inércia em relação ao eixo dos $x$ e $y$ são respetivamente ilustradas na figura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P7/MAI_Ch06_P7c.jpg"
width="600">

\begin{equation*}
\begin{aligned}
I_x &= 2\left(\frac{bh^3}{12}\right)
= \frac{(0.1)(0.08)^3}{6} = 8.533\times 10^{-6} ~\textrm{mm}^4\\
I_y &= 2\left(\frac{hb^3}{12}\right)
= \frac{(0.16)(0.05)^3}{6}
= 3.333\times 10^{-6} ~\textrm{mm}^4
\end{aligned}
\end{equation*}

O fator de 2 nas equações anteriores é para completar a figura do losango pela soma de dois triângulos. Finalmente, o momento polar da figura é
simplesmente calculado por:

\begin{equation*}
\begin{aligned}
J_0^\Diamond &= I_x + I_y = 1.187\times 10^{-5} ~\textrm{m}^4
\end{aligned}
\end{equation*}

```python
import sympy as sy
B = 0.100 # unit: m
H = 0.160

b, h = sy.symbols('b h') # unit: mm
print('-------------')
print('Area Losango: ')
print('-------------')
IxC1 = 2*b*h**3/12
print('Ix (1) = ',IxC1)
IxC1n = IxC1.subs([(b,B),(h,H/2)])
print('       = ',IxC1n,' m4')
IyC1 = 2*h*b**3/12
print('Iy (1) = ',IyC1)
IyC1n = IyC1.subs([(b,B/2),(h,H)])
print('       = ',IyC1n,' m4')
J01 = IxC1 + IyC1
print('J0 (1) = ',J01)
J01n = IxC1n + IyC1n
print('       = ',J01n,' m4')
```

**Retângulo 1**

Analisando a figura geométrica do retângulo por consulta direta da tabela, é possível implementar uma das duas vias:

- (i) calcular os momentos de inércia em relação aos dois eixos $x$ e $y$ e, posteriormente, calcular o momento polar (nota: atender aos eixos relativos aos
quais as fórmulas são definidas)

- (ii) calcular momento polar em relação à origem diretamente por formulário fornecido na tabela (nota: considerar o ponto sobre o qual o momento polar é forncedido na tabela).

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P7/MAI_Ch06_P7d.jpg"
width="600">

\begin{equation*}
\begin{aligned}
J_C^\blacksquare \equiv J_0^\blacksquare &= \frac{1}{12}bh(b^2 + h^2)
\Big|_{b=h}
= \frac{b^4}{6} = 4.267\times 10^{-7} ~\textrm{m}^4
\end{aligned}
\end{equation*}

```python
B2 = 0.040
J02 = B2**4/6
print('---------------------')
print('Área 2 (retângulo 1): ')
print('---------------------')
print('J0 (2) = ',J02,' m4')
```

**Retângulo 2**

De igual forma para o segundo retângulo, pode-se calcular o momento polar diretamente pela tabela. Contudo, de notar que neste caso é necessário aplicar o teorema dos eixos paralelos uma vez que o momento polar é dado em relação a um ponto $C$, centroide da figura, não coincidente com o ponto $O$
da origem do referencial considerado:


\begin{equation*}
J_0^\square = J_C^\square|_{\textrm{tabela}} + A\,
d^2
\end{equation*}

em que $d$ é a distância entre eixos que passam em $C$ e $0$.

\begin{equation*}
J_0^\square = \frac{1}{12}bh(b^2 + h^2)\Big|_{b=h} + Ad^2
= \frac{b^4}{6} + Ad^2
= 3.867\times 10^{-7} ~\textrm{m}^4
\end{equation*}

```python
B3 = 0.020
A3 = B3*B3
print('---------------------')
print('Área 3 (retângulo 2): ')
print('---------------------')
JC3 = B3**4/6
print('JC3 (3) = ','{:6.3e}'.format(JC3),' m4')
print('A3 (3)  = ','{:6.3e}'.format(A3),' m2')
d3 = B2/2 + B3/2
print('d3 (3)  = ','{:6.3e}'.format(d3),' m')
J03 = JC3 + A3*d3**2
print(f'J0 (3)  = {J03:6.3e} m4')
```

Pelo que o momento polar de inércia da superfície em relação à origem do referencial, $J_0$, é determinado pela combinação:

\begin{equation*}
J_0 = J_0^\Diamond - J_0^\blacksquare - J_0^\square =
1.105 \times 10^{-5} ~\textrm{m}^4
\end{equation*}

```python
J0 = J01n - J02 - J03
print(f'J0 (1-2-3) = {J0:.4e} m4')
```

#### b) Considere agora uma placa fina com a superfície indicada, com espessura constante $t =$ 10 mm e massa específica $\rho$ = 7850 kg/m $^3$. Determine o momento de inércia da placa em relação ao eixo $z$. <a id="B"></a> ([_enunciatum_](#P7))

Para uma placa delgada de espessura uniforme $t$, feita de um material homogêneo de massa específica $\rho$, o momento de inércia em relação a um eixo AA’ contido no plano da placa é:

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch06/P7/MAI_Ch06_P7e.jpg"
width="600">

\begin{equation*}
J_{AA', \textrm{massa}} = \int r^2 \mathrm{d}m \Bigg|_{\mathrm{d}m = \rho t
\mathrm{d}A}
= \rho t  \int r^2 \mathrm{d}A
= \rho t  I_{AA', \textrm{area}}
\end{equation*}

De modo semelhante, para um eixo BB’ contido no plano da placa e perpendicular a AA’, tem-se

\begin{equation*}
J_{BB', \textrm{massa}}
= \rho t  \int r^2 \mathrm{d}A
= \rho t  I_{BB', \textrm{area}}
\end{equation*}

Considerando agora o eixo CC’ perpendicular ao plano de placa e passando pelo ponto de intersecção C de AA’ e BB’, escreve-se:

\begin{equation*}
J_{CC', \textrm{massa}}
= \rho t  \underbrace{I_{C, \textrm{area}}}_{J_C}
= \rho t  (I_{AA', \textrm{area}} + I_{BB', \textrm{area}})
\end{equation*}

O momento de segunda ordem de área polar da placa em relação ao eixo $z$ é:

\begin{equation*}
J_z = \rho t \underbrace{(I_{x, \textrm{area}}
+ I_{y, \textrm{area}})}_{J_0}
\end{equation*}

Com $\rho = 7850 $ kg/m$^3$ e $t=0.01$ m, resulta:

\begin{equation*}
J_z = \rho t J_0
= 8.677\times 10^{-4} \quad\textrm{m$^4$}
\end{equation*}

```python
rho = 7850
t = 10e-3 # unit: m
Jz = rho*t*J0
print(f'J0 = {Jz:.3e} m4')
```
