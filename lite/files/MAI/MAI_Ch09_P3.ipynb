```python
L = 1 # assumir e.g. por unidade de comprimento
W = 98.1 # unit: N
F = 25.0
```

**Movimento virtual admissível, $\delta\theta$**

Considere-se um sistema de coordenadas $S(D,\vec{i},\vec{j})$, com origem em D. De notar que o ponto D é um ponto fixo no mecanismo.

Considere-se agora um movimento virtual do mecanismo. Esta perturbação do sistema de equilíbrio pode ser qualquer embora cinematicamente admissível, i e., respeitando as restrições sobre graus de liberdade ou deslocamentos prescritos do mecanismo.

A análise do sistema permite concluir que este é um problema a 1 grau de liberdade, pois o movimento virtual é completamente descrito pela variação da rotação medida pelo ângulo $\theta$ da barra DC.

**Identificação das forças externas que realizam trabalho virtual**

Assuma-se uma rotação virtual $\delta\theta$ da barra DC, em relação à sua configuração real inicial. Desenhe-se a configuração deformada virtual do sistema conforme se ilustra na figura abaixo. Esta configuração virtual deverá ser cinematicamente admissível (respeitar as condições de fronteira em deslocamento/rotação).

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch09/P3/MAI_Ch09_P3b.jpg"
style="max-width: 100%;"/>

Confrontando o sistema real com a configuração virtual identificam-se as forças no sistema real que produzirão um trabalho virtual (dado por um produto escalar entre vetores):

- As forças do peso próprio das duas barras, $P$, aplicadas nos seus centroides;
- A força horizontal diretamente aplicada em B, $F$.

De notar ainda que:

- Em D, as forças de reações segundo $x$ ($D_x$) e $y$ ($D_y$) não realizam trabalho virtual porque não existem quaisquer deslocamentos associados (apoio duplo).

- Em B, o deslocamento virtual que lhe está associado ($\delta x_B$) é perpendicular à direção da força de reação $B_y$ pelo que, também não realiza qualquer trabalho uma vez que o trabalho realizado por uma força segundo um
deslocamento que lhe é perpendicular é nulo (produto escalar entre vetores com linha de ação perpendiculares).

**Princípio dos Trabalhos Virtuais**

Por aplicação do princípio dos trabalhos virtuais:

\begin{equation*}
\delta U (\delta\theta) = \delta U_F + \delta U_{P} + \delta U_{P} = 0
~\Leftrightarrow~
\vec{F}\cdot\vec{\delta}_B +
\vec{P}\cdot\vec{\delta}_P +
\vec{P}\cdot\vec{\delta}_P = 0
\end{equation*}

em que,

- $\delta U_F$ é o trabalho virtual que a força $\vec{F}$ realiza no deslocamento virtual $\vec{\delta}_B$ do seu ponto de aplicação em B;

- $\delta U_P$ é o trabalho virtual que a força do peso próprio $\vec{P}$ realiza no deslocamento virtual $\vec{\delta}_P$ do seu ponto de aplicação no centroide da barra;

_Vetores força_

\begin{equation*}
\vec{F} = F \,\vec{\imath}
\quad\wedge\quad
\vec{P} = -P \,\vec{\jmath}
\end{equation*}

- Deslocamento virtual $\vec{\delta}_B$

O vetor deslocamento virtual em B tem apenas uma componente cartesiana horizontal devida ao apoio simples: $\vec{\delta}_B = x_B \vec{\imath}$.

De relações geométricas, considere-se a seguinte expressão trigonométrica (ver figura):

\begin{equation*}
x_B = 2L\cos\theta
\end{equation*}

Uma perturbação ou movimento virtual $\delta\theta$ irá provocar um deslocamento virtual $\delta x_B$ dada derivação:

\begin{equation*}
\delta x_B = -2L\sin\theta\,\delta\theta
\quad\therefore\quad
\vec{\delta}_B = \left(-2L\sin\theta\delta\theta\right)\vec{\imath}
\end{equation*}

De notar que o sentido negativo, indica que uma inversão de sentidos em relação ao sentido positivo para a rotação.

_Deslocamentos virtuais (pontos de aplicação das forças)_

- Deslocamento virtual $\vec{\delta}_P$

O vetor deslocamento virtual em P tem componentes cartesianas, $\vec{\delta}_P = x_P \vec{\imath} + y_P \vec{\jmath}$:

\begin{equation*}
\begin{cases}
x_P = x_B / 2 \\
y_P = - (L/2) \sin\theta
\end{cases}
\end{equation*}

por derivação:

\begin{equation*}
\begin{cases}
\delta x_P= \delta x_B / 2 = -L\sin\theta\delta\theta\\
\delta y_P = - (L/2) \cos\theta\delta\theta
\end{cases}
\quad\therefore\quad
\vec{\delta}_P =
(-L\sin\theta\delta\theta)\vec{\imath} +
\left(-\frac{L}{2}\cos\theta\delta\theta\right)
\vec{\jmath}
\end{equation*}

De notar, contudo, que o desenvolvimento do produto escalar $\vec{P}\cdot\vec{\delta}_P$ apenas envolve a componente vertical ($\delta y_P$).

- PTV (desenvolvendo o produto escalar)

\begin{equation*}
\begin{aligned}
\delta U = 0
&~\Leftrightarrow~
F\cdot\delta x_B +
P\cdot\delta y_P +
P\cdot\delta y_P = 0\\
&~\Leftrightarrow~
F\cdot (-2L\sin\theta\,\delta\theta)
+ 2(-P) \cdot (-\frac{L}{2}\cos\theta\,\delta\theta) = 0 \\
&~\Leftrightarrow~
-2LF\sin\theta + P L \cos\theta = 0
\quad\wedge\quad \delta\theta \ne 0 \\
&~\Leftrightarrow~
\tan\theta \equiv\frac{\sin\theta}{\cos\theta} =
\frac{P}{2F}
\end{aligned}
\end{equation*}

```python
import numpy as np
theta = np.arctan(W/(2*F))
print(f' theta = {theta:.3f} radianos = {np.rad2deg(theta):.1f} graus')
```

```python
import matplotlib.pylab as plt
from ipywidgets import interactive


# Define a switch for the color theme
use_light_theme = True

if use_light_theme:
    # Solarized Light Theme Colors
    background_color = (253/255, 246/255, 227/255)
    font_color = (0.1, 0.1, 0.1)
    grid_color = (147/255, 161/255, 161/255)
else:
    # Dark Theme Colors
    background_color = (.15, .15, .15)
    font_color = (.8, .8, .8)
    grid_color = (.2, .2, .2)

plt.rcParams['axes.facecolor'] = background_color
plt.rcParams['figure.facecolor'] = background_color
plt.rcParams['font.family'] = 'monospace'
plt.rcParams['font.size'] = 18
# plt.rcParams['text.usetex'] = True

params = {
    "ytick.color": font_color,
    "xtick.color": font_color,
    "grid.color": grid_color,
    "text.color": font_color,
    "axes.labelcolor": font_color,
    "axes.edgecolor": (.15, .15, .15)
}

plt.rcParams.update(params)

vart = np.arange(0,91,5)

def mecanismo(vart):
    xB, xBeq = 2*L*np.cos(np.deg2rad(vart)), 2*L*np.cos(theta)
    xP, xPeq = xB/2, xBeq/2
    yP, yPeq = -L/2*np.sin(np.deg2rad(vart)), -L/2*np.sin(theta)
    plt.figure(figsize=(6,4))
    xx, yy = [0,xP,xB], [0,yP,0]
    plt.plot(xx,yy,'-')
    # posição de equilibrio
    xxeq, yyeq = [0,xPeq,xBeq], [0,yPeq,0]
    plt.plot(xxeq,yyeq,'--')
    plt.xlim(0,2*L)
    plt.show()

interactive_plot = interactive(mecanismo, vart=(0, 90))
output = interactive_plot.children[-1]
output.layout.height = '250px'
interactive_plot
```
