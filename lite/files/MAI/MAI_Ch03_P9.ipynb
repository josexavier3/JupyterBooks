### Resolução

#### a) Calcule justificadamente o comprimento mínimo da barra para o qual esta permanece em equilíbrio estático.

O comprimento da barra tem influência sobre a linha de ação da força $P$ relativa ao efeito do seu peso próprio. Como se procura ilustrar na Fig. (a i) se a barra for relativamente curta (comprimento mínimo) existirá uma tendência para esta rodar no sentido anti-horário, sobre o ponto $B$, deslizando sobre a superfície vertical no ponto $A$. Este cenário introduzirá forças de atrito, no ponto $A$ e $B$ que tenderão a contrariar esse movimento de rotação. A condição de equilíbrio ocorrerá precisamente na condição do comprimento mínimo da barra que equilibra as forças externas representadas no diagrama de corpo livre (DCL) da Fig. (a.ii).

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P9/MAI_Ch03_P9a.jpg"
width="900">

Analisando o equilíbrio das forças com componentes escalares de acordo com um sistema de coordenadas escolhido:

\begin{equation*}
\begin{cases}
\rightarrow(+)~&\sum F_x &= &0 \\
\uparrow(+)~&\sum F_y &= &0 \\
\circlearrowleft(+)~&\sum M_A &= &0
\end{cases}~\Leftrightarrow~
\begin{cases}
N_A - N_B\sin\theta + F_{a_B}\cos\theta = 0 \\
F_{a_A} - P + N_B\cos\theta + F_{a_B}\sin\theta = 0 \\
-P\cdot\displaystyle\frac{l}{2}\cos\theta + N_B \cos\theta \cdot d
+ N_B \sin\theta \cdot (d\tan\theta)
+ F_{a_B}\sin\theta\cdot d
- F_{a_B}\cos\theta\cdot (d\tan\theta)  = 0
\end{cases}
\end{equation*}

ou, simplificadamente, considerando os vetores braço ($\vec{r}$) e força ($\vec{F}$) perpendiculares entre si, no desenvolvimento da expressão do equilíbrio do momento (de notar que nesta abordagem o vetor $\vec{F}_{aA}$ não exerce momento em relação ao ponto $A$, uma vez que a sua linha de ação contem esse ponto),

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0 \\
\sum M_A = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
N_A - N_B\sin\theta + F_{a_B}\cos\theta = 0 \\
F_{a_A} - P + N_B\cos\theta + F_{a_B}\sin\theta = 0 \\
-P\cdot\displaystyle\frac{l}{2}\cos\theta + N_B\cdot H  = 0
\end{cases}
\end{equation*}

em que,

\begin{equation*}
\theta = 30^\circ\frac{\pi}{180^\circ}~\textrm{(radianos)}~\therefore~
\cos\theta = \frac{\sqrt{3}}{2}
\quad\textrm{e}\quad
H = \frac{d}{\cos\theta} = \frac{4}{\sqrt{3}}
\end{equation*}

De notar ainda que (no limiar do início do movimento),

\begin{equation*}
\begin{cases}
F_{a_A} &=& \mu_e N_A \\
F_{a_B} &=& \mu_e N_B \\
P &=& mg  \\
\sin\theta\cdot d &=& \cos\theta\cdot d\tan\theta
\end{cases}
\end{equation*}

Pelo que resulta,

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0 \\
\sum M_A = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
N_A - N_B\sin\theta + \mu_e N_B\cos\theta = 0 \\
\mu_e N_A - P + N_B\cos\theta + \mu_e N_B\sin\theta = 0  \\
-P \displaystyle\frac{l}{2}\cos\theta
+ N_B \frac{4}{\sqrt{3}} = 0
\end{cases}\Leftrightarrow~
\begin{cases}
N_A + N_B\left(-\displaystyle\frac{1}{2} + \mu_e
\displaystyle\frac{\sqrt{3}}{2}\right) = 0 \\
\mu_e N_A - P + N_B\left(\displaystyle\frac{\sqrt{3}}{2}
+ \mu_e \displaystyle\frac{1}{2}\right)= 0 \\
-\displaystyle\frac{\sqrt{3}}{4}Pl
+ N_B \displaystyle\frac{4}{\sqrt{3}} = 0
\end{cases}
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

m = 10   # unit: kg
g = 9.81 # unit: m/s2
P = m*g
print('P   = ','{:.2f}'.format(P),' N')
d = 2   # unit: m
mue = 0.4
ang30 = np.deg2rad(30) # unit: radianos
ang60 = np.deg2rad(60)

na, nb, l = sy.symbols('na nb l')

faa = mue*na
fab = mue*nb
print('-------------------')
print('Sistema de equações')
print('-------------------')
sumFx = na - nb*np.sin(ang30) + fab*np.cos(ang30)
print('Sum Fx :',sumFx)
sumFy = faa - P + nb*np.cos(ang30) + fab*np.sin(ang30)
print('Sum Fy :',sumFy)
sumMA = - P*(l/2)*np.cos(ang30) + nb*(d/np.cos(ang30))
print('Sum sumMA :',sumMA)
sol = solve({sumFx,sumFy, sumMA},{na,nb,l})
print('-------------------')
print('Solução')
print('-------------------')
NA = sol[na]
print('NA  = ','{:.3f}'.format(NA),' N')
NB = sol[nb]
print('NB  = ','{:.3f}'.format(NB),' N')
L = sol[l]
print('l   = ','{:.3f}'.format(L),' m')
print('-------------------')
print('+ F atrito')
print('-------------------')
print('FaA = ','{:.3f}'.format(faa.subs(na,NA)),' N')
print('FaB = ','{:.3f}'.format(fab.subs(nb,NB)),' N')
```

Resolvendo na forma matricial,

\begin{equation*}
[A]\{x\} = \{b\} \Leftrightarrow
\begin{bmatrix}
1 & \left(-\displaystyle\frac{1}{2} + \mu_e
\displaystyle\frac{\sqrt{3}}{2}\right) & 0 \\
\mu_e & \left(\displaystyle\frac{\sqrt{3}}{2}
+ \mu_e \displaystyle\frac{1}{2}\right) & 0\\
0 & \displaystyle\frac{4}{\sqrt{3}} & -\displaystyle\frac{\sqrt{3}}{4}P
\end{bmatrix}
\begin{bmatrix}
N_A \\ N_B \\ l
\end{bmatrix} =
\begin{bmatrix}
0 \\ P \\ 0
\end{bmatrix}
\end{equation*}

```python
import numpy as np
np. set_printoptions(precision=3, suppress=True)

A0 = [1, -1/2+mue*np.sqrt(3)/2, 0]
A1 = [mue, np.sqrt(3)/2+1/2*mue, 0]
A2 = [0, 4/np.sqrt(3), -np.sqrt(3)/4*P]

A = np.array([A0,A1,A2])
print(f'A =\n {A}')

invA = np.linalg.inv(A)
print(f'A^-1 =\n {invA}')

b = np.array([0,P,0])
print(f'b = {b}')

x = np.linalg.solve(A,b)
print('-------------------')
print('Solução')
print('-------------------')
print(f'x = [NA NB l] = {x} [N - m]')
print('+ F atrito: -------')
print('FaA = ','{:.3f}'.format(x[0]*mue),' N')
print('FaB = ','{:.3f}'.format(x[1]*mue),' N')
```

#### b) Sem efetuar cálculos, explique como calcularia o comprimento máximo da barra de forma a permanecer em repouso. Represente o respectivo diagrama de corpo livre.

Para o comprimento máximo, a barra iria ter tendência a descair para a direita devido à ação do peso neste sentido, como esquematicamente representado na Fig. (b.i). Neste caso, as forças de atrito que sugeriam nos pontos de apoio $A$ e $B$ seriam no sentido de contrariar este movimento. Na Fig. (b.ii) representa-se o diagrama de corpo livre da barra onde se indica o sentido das forças de reação (normal e atrito).

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch03/P9/MAI_Ch03_P9b.jpg"
width="1000">

Desta forma, a análise do equilíbrio do DCL da figura (b.ii) conduz ao seguinte sistema de equações:

\begin{equation*}
\begin{cases}
\sum F_x = 0 \\
\sum F_y = 0 \\
\sum M_A = 0
\end{cases}~\Leftrightarrow~
\begin{cases}
N_A - N_B\sin\theta - \mu_e N_B\cos\theta = 0 \\
-\mu_e N_A - P + N_B\cos\theta - \mu_e N_B\sin\theta = 0  \\
-P \displaystyle\frac{l}{2}\cos\theta
+ N_B \frac{4}{\sqrt{3}} = 0
\end{cases}
\end{equation*}

A resolução do sistema de equações conduz à solução:

```python
print('-------------------')
print('Sistema de equações')
print('-------------------')
sumFxb = na - nb*np.sin(ang30) - fab*np.cos(ang30)
print('Sum Fx :',sumFxb)
sumFyb = -faa - P + nb*np.cos(ang30) - fab*np.sin(ang30)
print('Sum Fy :',sumFyb)
sumMAb = - P*(l/2)*np.cos(ang30) + nb*(d/np.cos(ang30))
print('Sum sumMA :',sumMAb)
solb = solve({sumFxb,sumFyb, sumMAb},{na,nb,l})
print('-------------------')
print('Solução')
print('-------------------')
NAb = solb[na]
print('NA  = ','{:.3f}'.format(NAb),' N')
NBb = solb[nb]
print('NB  = ','{:.3f}'.format(NBb),' N')
Lb = solb[l]
print('l max.  = ','{:.3f}'.format(Lb),' m')
print('-------------------')
print('+ F atrito')
print('-------------------')
print('FaA = ','{:.3f}'.format(faa.subs(na,NAb)),' N')
print('FaB = ','{:.3f}'.format(fab.subs(nb,NBb)),' N')
```
