## CENTROIDES E CENTROS DE GRAVIDADE

### Problema 1

Considere uma peça construída em chapa de aço com a forma indicada na figura. A peça é simétrica em relação ao plano paralelo a $xz$ que passa nos pontos H e D, as linhas AC, EG e GA são segmentos de recta, a linha CDE é uma semi-elipse e o furo de centro em O é circular. Determine a posição do centróide da área da superfície superior da peça.

### Resumo

Em vários casos é possível decompor uma placa plana em geometrias simples, como retângulos, triângulos e círculos. Desta forma, é possível determinar a posição ou coordenadas do centro de gravidade (centro de massa e centroide), $(\overline{X},\overline{Y})$ a partir das coordenadas dos centros de gravidade de cada subárea que o constituem, $\{(\overline{x}_1,\overline{y}_1),(\overline{x}_2,\overline{y}_2),\ldots,(\overline{x}_n,\overline{y}_n)\}$. O centro de gravidade de uma placa composta é definido pelas seguintes expressões:

\begin{equation*}
\overline{X} = \frac{\sum_i \overline{x}_i P_i}{P}
\qquad
\overline{Y} = \frac{\sum_i \overline{y}_i P_i}{P}
\end{equation*}

Se a placa for homogénea e de espessura uniforme, então o centro de gravidade coincide como centroide da secção. Neste caso substituindo:

\begin{equation*}
P = \gamma t A
\end{equation*}

- $\gamma$-peso específico do material [N/m $^3$]
- $t$-espessura da placa [m]
- $A$-área da superfície do elemento, [m $^2$]
  
obtêm-se as seguintes expressões para o momento estático ou momento de 1ª ordem da superficie A relativamente ao eixo $y$ e $x$:

\begin{equation*}
Q_\textrm{y} = \overline{X}\,\sum A = \sum\overline{x}\,A
\qquad
Q_\textrm{x} = \overline{Y}\,\sum A = \sum\overline{y}\,A
\end{equation*}

### Resolução

A metodologia de resolução do problema passar por dividir a área da placa em componentes (subáreas) simples, encontrar o centroide de cada componente elementar (ver formulário para áreas comuns), determinar os primeiros momentos de área segundo os eixos $x$ e $y$ e finalmente obter as coordenadas do centroide ou centro geométrico.

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/Ch05/P1/MAI_Ch05_P1a.jpg"
width="800">

A posição do centroide da peça pode ser determinada recorrendo às expressões:

\begin{equation*}
\overline{X} = \displaystyle\frac{Q_y}{\displaystyle\sum_{i=1}^n A_i} =
\displaystyle\frac{\displaystyle\sum_{i=1}^3
\overline{x}_iA_i}{\displaystyle\sum_{i=1}^3 A_i}
= \displaystyle\frac{\overline{x}_1A_1 + \overline{x}_2A_2 -
\overline{x}_3A_3}{A_1+A_2-A_3}
\end{equation*}

\begin{equation*}
\overline{Y} = \displaystyle\frac{Q_x}{\displaystyle\sum_{i=1}^n A_i} =
\displaystyle\frac{\displaystyle\sum_{i=1}^3
\overline{y}_iA_i}{\displaystyle\sum_{i=1}^3 A_i}
= \displaystyle\frac{\overline{y}_1A_1 + \overline{y}_2A_2 -
\overline{y}_3A_3}{A_1+A_2-A_3}
\end{equation*}

de onde,

| SubÁreas|A, mm $^2$|$\overline{x}$, mm|$\overline{y}$, mm|$\overline{x}A$, mm $^3$|$\overline{y}A$, mm $^3$|
|:-------:|:-------:|:---------------:|:---------------:|:--------------------:|:--------------------:|
| (+)1 | 4800.0 | 30.0 | 40.0 | 144000.0 | 192000.0 |
| (+)2 | 1885.0 | 72.7 | 40.0 | 137097.3  | 75398.2 |
| (-)3 | 78.5 | 60.0  | 40.0 | 4712.4 | 3141.6 |
| $\sum$ | 6606.4 | - | - | $Q_y =$ 276384.9   | $Q_x =$ 264256.6  |

```python
import numpy as np

# parametros, # unit: mm
AB, BC, CD, AH, HG = 30, 30, 30, 40, 40
diam = 10

# Area 1:
A1 = (AB+BC)*(AH + HG)
x1, y1 = AB, AH
print('Area 1: ------- ')
print(f'A1 = {A1:.1f} mm2 | x1 = {x1:.1f} mm | y1 = {y1:.1f} mm')
print(f'x1.A1 = {x1*A1:.1f} mm3')
print(f'y1.A1 = {y1*A1:.1f} mm3')
```

```python
# Area 2:
A2 = np.pi*AH*CD/2
print('Area 2: ------- ')
x2 = (AB+BC) + 4*CD/3/np.pi 
y2 = AH 
print(f'A2 = {A2:.1f} mm2 | x2 = {x2:.1f} mm | y2 = {y2:.1f} mm')
print(f'x2.A2 = {x2*A2:.1f} mm3')
print(f'y2.A2 = {y2*A2:.1f} mm3')
```

```python
# Area 3:
A3 = np.pi*(diam/2)**2
x3, y3 = (AB+BC), AH  
print('Area 3: ------- ')
print(f'A3 = {A3:.1f} mm2 | x3 = {x3:.1f} mm | y3 = {y3:.1f} mm')
print(f'x3A2 = {x3*A3:.1f} mm3')
print(f'y3A2 = {y3*A3:.1f} mm3')
```

```python
print('Area total : ------- ')
At = A1 + A2 - A3
print('A = ','{:.1f}'.format(At),' mm2')
Qy = A1*x1 + A2*x2 - A3*x3
print('Momento 1º ordem: Qy = ','{:.1f}'.format(Qy),' mm3')
Qx = A1*y1 + A2*y2 - A3*y3
print('Momento 1º ordem: Qx = ','{:.1f}'.format(Qx),' mm3')

print('Centroides : ------- ')
X_centr = Qy/At
print('X_centr = ','{:.1f}'.format(X_centr),' mm')
Y_centr = Qx/At # Y_centr = AH: pelo eixo de simetria da peca
print('Y_centr = ','{:.1f}'.format(Y_centr),' mm')
```

É importate notar que, mesmo sem efetuar cálculos, a posição do centroide pode ser parcialmente identificada atendendo a que existência de um eixo de simetria, pelo que o centroide ficará situado sobre esse eixo. Por conseguinte, observando que a linha HD corresponde a um eixo de simetria (segundo o eixo $x$) da peça a coordenada em $y$ do centróide é: $\overline{y} = 40$ mm.


### Apêndice

#### Corpo

Considere-se um corpo como um sistema infinito de pontos materiais de massa infinitesimal.

#### Centro de gravidade

É o ponto sobre o qual a aplicação de uma única força resultante é equivalente ao efeito gravitacional sobre cada uma das partículas que formam um corpo rígido. Se o campo gravitacional for homogéneo ($g:$ constante), o centro de gravidade coincide com o centro de massa.

No estudo do equilíbrio de um corpo rígido mostrou-se como um sistema de forças genérico é equivalente a um sistema força-binário. Nesta secção, estendemos essa ideia para mostrar como um sistema de forças (em particular, elementos do peso próprio de um objeto) pode ser substituído por uma única força resultante aplicada num ponto específico (a resultante dos momentos das forças em relação a esse ponto é nulo). O ponto específico é chamado de centro de gravidade do objeto.

**Resultante do peso e centro de gravidade de um sólido plano**

\begin{equation*}
P = \int \mathrm{d}P~; \quad
\overline{x}\,P = \int x\,\mathrm{d}P~; \quad
\overline{y}\,P = \int y\,\mathrm{d}P
\end{equation*}

#### Centro de massa

O centro de massa é a média aritmética de todas os pontos ponderados pelo peso específico ($\gamma$, N/m$^3$) ou massa específica ($\rho$, kg/m$^3$, com $\gamma = \rho g$, $g=9.81$ m/s$^2$) do material. Se um objecto tem densidade uniforme, o seu centro de massa coincide com o centroide (da sua forma) : $P = \gamma V = \gamma t A$ (placa homogénea de espessura uniforme).

#### Centroide

Ponto ou coordenada de uma forma geométrica que estabelece o seu centro
geométrico. Em geometria, baricentro é sinónimo de centroide. O centroide
(centro geométrico do corpo) é coincidente com o centro de massa se o corpo
tiver massa específica uniforme.

**Centroide de uma superfície**

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/GFormulas/CentroidArea.jpg"
width="600">

\begin{equation*}
\begin{aligned}
\overline{x}\,A &= \int x\,\mathrm{d}A = Q_y \quad (\textrm{momento de 1º ordem
da superfície em relação ao eixo $y$}) \\
\overline{y}\,A &= \int y\,\mathrm{d}A = Q_x \quad (\textrm{momento de 1º ordem
da superfície em relação ao eixo $x$})
\end{aligned}
\end{equation*}

Os primeiros momentos da área são úteis na mecânica dos sólidos deformáveis
para determinar as tensões de corte em vigas submetidas a cargas
transversais ao seu eixo.

**Centroide de uma linha**

<img src="https://userweb.fct.unl.pt/~jmc.xavier//MAI/Notebooks/GFormulas/CentroidLine.jpg"
width="600">

Para uma linha homogénea e de secção transversal uniforme: $P=\gamma S L$, em
que $S -$ é área de secção transversal do fio, $L -$ é o comprimento do elemento
do fio

\begin{equation*}
\begin{aligned}
\overline{x}\,L &= \int x\,\mathrm{d}L \\
\overline{y}\,L &= \int y\,\mathrm{d}L
\end{aligned}
\end{equation*}



#### Determinação da posição do centroide

- Se existir um eixo de simetria, o centroide ficará situado sobre esse eixo.

- Se existirem dois ou mais eixos de simetria, o centroide ficará situado no ponto onde se intercetam.

- Se existir um centro de simetria, o centroide coincidirá com esse centro.
