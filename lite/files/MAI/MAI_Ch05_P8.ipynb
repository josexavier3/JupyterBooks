### Resolução

#### a) Determine, efetuando os cálculos por integração, o momento estático da superfície representada em relação ao eixo $y$. (Nota 1: recorde a propriedade aditiva da integração). ([_enunciatum_](#P8))

A superfície enunciada por ser obtida subtraindo a um triângulo um setor circular, resultando na decomposição em figuras simples ilustradas na figura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch05/P8/MAI_Ch05_P8a.jpg"
style="max-width: 100%;"/>

O cálculo por integração do momento estático da superfície em relação ao eixo $y$, pode ser obtido por partes, i.e. subtraindo o resultado do cálculo por integração da área do triângulo com o cálculo pelo integral do setor circular, ambos em relação ao eixo $y$.

**Triângulo**

No caso do triângulo, começa-se por definir as retas $y_1(x)$ e $y_2(x)$ conforme se ilustra na figura:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch05/P8/MAI_Ch05_P8c.jpg"
style="max-width: 100%;"/>

O declive de ambas as retas pode ser diretamente expresso considerando a tangente do ângulo de 30$^\circ$, sendo a primeira ($y_1(x)$) positiva e a segunda ($y_2(x)$) negativa, ou seja:

\begin{equation}
y_1(x) = \frac{y_a}{a} x \Bigg|_{(x_0,y_0)= (0,0)}= \tan(\theta)\,
x\Bigg|_{\theta =
30^\circ} = \frac{\sqrt{3}}{3} x
\end{equation}

e

\begin{equation}
y_2(x) = - y_1(x) = - \frac{\sqrt{3}}{3} x
\end{equation}

_Área :_

Área do triângulo vem:

\begin{equation}
A = \int\,\mathrm{d}A
= \int_{\large 0}^{\large a}\int_{\large-\sqrt{3}x/3}^{\large\sqrt{3}x/3}
\mathrm{d}y\mathrm{d}x
\Bigg|_{\begin{aligned} &\textrm{resolvendo primeiro o
integral} \\
&\textrm{que não tem os limites fixos}
\end{aligned}}
\end{equation}

\begin{equation}
\begin{aligned}
~~& = \int_{\large 0}^{\large a}
[y]_{\large-\sqrt{3}x/3}^{\large\sqrt{3}x/3}
\mathrm{d}x
= 2\frac{\sqrt{3}}{3}\int_{\large 0}^{\large a}
x \mathrm{d}x
= 2\frac{\sqrt{3}}{3}
\left[\frac{x^2}{2}\right]_{\large 0}^{\large a}
= \frac{\sqrt{3}}{3} a^2 = 0.577 a^2 \quad\textrm{[m$^2$]}
\end{aligned}
\end{equation}

```python
import numpy as np
import sympy as sy

a, x, y = sy.symbols('a x y')

xi = 0 # limits on x
xf = a
yi = -np.sqrt(3)/3*x # limits on y
yf = np.sqrt(3)/3*x

f = 1 # integrant function
A1 = sy.integrate(f, (y,yi,yf), (x,xi,xf))
print('A1 =',sy.factor(A1),' m2')
```

_Momento estático :_

Por sua vez, o momento estático ou de primeira ordem da área triangular em relação ao eixo dos $y$ é:

\begin{equation}
\begin{aligned}
Q_y = \int x\,\mathrm{d}A
&=  \int_{\large 0}^{\large a}
x\left(\int_{\large-\sqrt{3}x/3}^{\large\sqrt{3}x/3}
\mathrm{d}y\right)\mathrm{d}x
\\
~~&= \int_{\large 0}^{\large a}
x[y]_{\large-\sqrt{3}x/3}^{\large\sqrt{3}x/3}
\mathrm{d}x
= 2\frac{\sqrt{3}}{3}\int_{\large 0}^{\large a}
x^2 \mathrm{d}x \\
&= 2\frac{\sqrt{3}}{3}
\left[\frac{x^3}{3}\right]_{\large 0}^{\large a}
= \frac{2\sqrt{3}}{9} a^3 = 0.385 a^3 \quad\textrm{[m$^3$]}
\end{aligned}
\end{equation}

```python
f = x # integrant function
Qy1 = sy.integrate(f, (y,yi,yf), (x,xi,xf))
print('Qy1 =',sy.factor(Qy1),' m3')
```

**Setor circular**

Analisando agora o cálculo por integração da área e do momento estático relativamente ao eixo $y$ do sector circular:

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MAI/Notebooks/Ch05/P8/MAI_Ch05_P8d.jpg"
style="max-width: 100%;"/>

_Área :_

Área do setor circular é:

\begin{equation}
\begin{aligned}
A &= \int\,\mathrm{d}A = \int\,r\mathrm{d}\theta\mathrm{d}r
= \int_{\large 0}^{\large R}r
\left(\int_{\large-\pi/6}^{\large\pi/6}
\mathrm{d}\theta\right)\mathrm{d}r
= \int_{\large 0}^{\large R}r
[\theta]_{\large-\pi/6}^{\large\pi/6}\mathrm{d}r \\
& = \frac{\pi}{3}\int_{\large 0}^{\large R}
r\,\mathrm{d}r = \frac{\pi}{3}\left[\frac{r^2}{2}\right]_{\large 0}^{\large R}
= \frac{\pi}{6}R^2
= 0.524 \,R^2 \quad\textrm{[m$^2$]}
\end{aligned}
\end{equation}

```python
r, t = sy.symbols('r t')

ri = 0 # limits on x
rf = r
ti = -np.pi/6
tf = np.pi/6

f = r # integrant function
A2 = sy.integrate(f, (t,ti,tf), (r,ri,rf))
print('A2 =',sy.factor(A2),' m2')
```

_Momento estático :_

Momento estático ou de primeira ordem da área triangular em relação ao eixo dos $y$:

\begin{equation}
\begin{aligned}
Q_y &= \int x\,\mathrm{d}A = \int x r \,\mathrm{d}\theta\mathrm{d}r =
\int_{\large 0}^{\large R}
\int_{\large-\pi/6}^{\large\pi/6}
\,(r\cos\theta)\,r\mathrm{d}\theta\mathrm{d}r
= \int_{\large 0}^{\large R}r^2
\left(\int_{\large-\pi/6}^{\large\pi/6} \cos\theta\,
\mathrm{d}\theta\right)\mathrm{d}r \\
&= \int_{\large 0}^{\large R}r^2
[\sin\theta]_{\large-\pi/6}^{\large\pi/6}\mathrm{d}r
= \int_{\large 0}^{\large R}r^2
[\sin(\pi/6)-\sin(-\pi/6)]\mathrm{d}r\Bigg|_{\sin(\pi/6) = 1/2}\\
& = \left[\frac{r^3}{3}\right]_{\large 0}^{\large R}
= \frac{1}{3}R^3 = 0.333 \,R^3 \quad\textrm{[m$^3$]}
\end{aligned}
\end{equation}

```python
f = r**2*sy.cos(t) # integrant function
Qy2 = sy.integrate(f, (t,ti,tf), (r,ri,rf))
print('Qy2 =',sy.factor(Qy2),' m3')
```

**Triângulo-setor circular**

A área da figura composta é:

\begin{equation}
\begin{aligned}
A = A^\lhd - A^\sphericalangle
&= \frac{\sqrt{3}}{3} a^2 - \frac{\pi}{6}R^2\Bigg|_{\large R = a/2}
= \frac{\sqrt{3}}{3} a^2 - \frac{\pi}{24}a^2 = 0.446 \,a^3 \quad\textrm{[m$^2$]}
\end{aligned}
\end{equation}

```python
A = A1 - A2.subs(r,a/2)
print('A = ',A,' m2')
```

Por sua vez, subtraindo ao momento estático do triângulo pelo momento estático do sector circular, obtém-se o momento estático (ou de 1ª ordem) da figura em relação ao eixo $y$:

\begin{equation}
\begin{aligned}
Q_y = Q_y^\lhd - Q_y^\sphericalangle
&= \frac{2\sqrt{3}}{9} a^3 - \frac{R^3}{3}\Bigg|_{\large R = a/2}
= \frac{2\sqrt{3}}{9} a^3 - \frac{1}{3}a^3 \\
&= \frac{1}{72} (-3 + 16 \sqrt{3}) a^3 =  0.343 \,a^3 \quad\textrm{[m$^3$]}
\end{aligned}
\end{equation}

```python
Qy = Qy1 - Qy2.subs(r,a/2)
print('Qy =',sy.factor(Qy),' m3')
```

Dinalmente, a coordenada do centroide da figura segundo o eixo $x$ é:

\begin{equation}
\begin{aligned}
\overline{X} = \frac{\displaystyle\int x \mathrm{d}A}{\displaystyle\int\mathrm{d}A}
= \frac{Q_y}{A} = \frac{(-3 + 16 \sqrt{3}) a^3/72}
{(\sqrt{3}/3)a^2 - (\pi/6)(a/2)^2}
= \frac{0.343 a^3}{0.446 a^2}
= 0.769 a
\end{aligned}
\end{equation}

Como $y=0$ o eixo $x$ é um eixo de simetria.

```python
x = Qy/A
print('x = ',x,' m')
```

#### b) Determine, justificando, as coordenadas do centroide da superfície no referencial $S(0,x,y)$ indicado. (Nota 2: pode utilizar a tabela que se encontra no verso). ([_enunciatum_](#P8))

Considera-se agora a solução do problema via decomposição em figuras elementares simples de acordo com formulário tabelado. Considere-se a decomposição da figura abaixo:

**Triângulo**

_Área:_

\begin{equation}
A = \frac{1}{2}\left(2\frac{\sqrt{3}}{3}a\right)a = \frac{\sqrt{3}}{3}a^2
\end{equation}

_Centroide:_

\begin{equation}
\overline{x}_1 = a- \frac{a}{3} = \frac{2a}{3}
\end{equation}

| Área: &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;| Centroide: &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|
|:---|:---|
|$A=\displaystyle\frac{\sqrt{3}}{3}a^2$|$\overline{x}_1=\displaystyle\frac{2a}{3}$|

**Setor circular**

_Área:_

\begin{equation}
A = \alpha r^2 \Bigg|_{\alpha = 30^\circ}=
\frac{\pi}{6} R^2 = \frac{\pi}{24} a^2
\end{equation}

_Centroide:_

\begin{equation}
\overline{x}_2 = \frac{2r \sin\alpha}{3\alpha}
= \frac{2R (1/2) }{3(\pi/6)} = \frac{2R}{\pi}
= \frac{a}{\pi}
\end{equation}

| Área: &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | Centroide: &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|
|:---|:---|
|$A=\displaystyle\frac{\pi}{24} a^2$|$\overline{x}_2=\displaystyle\frac{a}{\pi}$|

**Tabela final**

De onde resulta:

| Áreas:&nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|$A_i$, m$^2$|$\overline{x}_i$, m&nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|$\overline{x}_iA_i$, m$^3$&nbsp; &nbsp; &nbsp; &nbsp; &nbsp;|
|:-------:|:----------:|:-----------------:|:------------------------:|
| `1` | $\displaystyle\frac{\sqrt{3}}{3}a^2$ | $\displaystyle\frac{2a}{3}$ | $\displaystyle\frac{2 a^3}{3\sqrt{3}}$ |
| `2` | $\displaystyle\frac{\pi}{24} a^2$ | $\displaystyle\frac{a}{\pi}$ |$\displaystyle\frac{a^3}{24}$ |
| $\sum : (1) - (2)$ | $\displaystyle\frac{a^2 (8\sqrt{3} - \pi)}{24} $ | - |$\displaystyle\frac{(-3 + 16\sqrt{3}) a^3}{72}$ |

ou seja,

\begin{equation}
\begin{aligned}
\overline{X} = \frac{Q_y}{A}
= \frac{\displaystyle\sum_{i=1}^2
\overline{x}_iA_i}{\displaystyle\sum_{i=1}^2 A_i}
= \frac{\overline{x}_1A_1 - \overline{x}_2A_2}{A_1-A_2}
= \frac{0.343 a^3}{0.446 a^2}
= 0.769 a
\end{aligned}
\end{equation}

#### c) Sabendo que o volume obtido pela rotação completa da superfície em torno do eixo $y$ é de 100 cm$^3$, determine, justificando, o valor de $a$. ([_enunciatum_](#P8))

\begin{equation}
\begin{aligned}
V = &\int \mathrm{d}V = \int 2 \pi x \mathrm{d}A
= 2 \pi \, \overline{X}\, A
= 2 \pi (0.769 a) (0.446 a^2)
= 100 ~\textrm{cm$^3$}\\
\Leftrightarrow\quad &a = 3.593 ~\textrm{cm}
\end{aligned}
\end{equation}

nota: as outras duas raizes da equação cúbica em $a$ são números complexos, e por isso sem significado físico no contexto do problema.

```python
from sympy.solvers import solve

Vn = 100 # unit: cm3

EQ = Vn - 2*np.pi*x*A
sol = solve(EQ,a)
an = sol[0]
print('a = ','{:.3f}'.format(an),' cm')
```
