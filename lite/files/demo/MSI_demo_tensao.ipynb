## Tensões principais e direções principais

Dado um tensor das tensões $\boldsymbol{\sigma}$ (MPa), calcular as tensões
principais (valores próprios) e as direções principais (vetores próprios).

```python
import numpy as np

sigma = np.array([
    [ 50.0, 30.0,  0.0],
    [ 30.0, -20.0, 0.0],
    [  0.0,  0.0, 10.0],
])

principais, direcoes = np.linalg.eigh(sigma)
print('Tensoes principais (MPa):', np.round(principais, 2))
print('Direcoes principais (colunas):')
print(np.round(direcoes, 3))
```

## Círculo de Mohr (plano xy)

```python
import numpy as np
import matplotlib.pyplot as plt

sx, sy, txy = 50.0, -20.0, 30.0
c = (sx + sy) / 2
r = np.hypot((sx - sy) / 2, txy)

th = np.linspace(0, 2 * np.pi, 200)
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(c + r * np.cos(th), r * np.sin(th))
ax.plot([sx, sy], [txy, -txy], 'o-')
ax.axhline(0, lw=0.5, color='k'); ax.axvline(0, lw=0.5, color='k')
ax.set_aspect('equal'); ax.set_xlabel('sigma (MPa)'); ax.set_ylabel('tau (MPa)')
ax.set_title(f'Circulo de Mohr: centro={c:.1f}, raio={r:.1f}')
plt.show()
```

## Solução simbólica com SymPy

```python
import sympy as sp

sx, sy, txy, th = sp.symbols('sigma_x sigma_y tau_xy theta', real=True)
sigma_n = (sx + sy)/2 + (sx - sy)/2*sp.cos(2*th) + txy*sp.sin(2*th)
th_p = sp.solve(sp.diff(sigma_n, th), th)
print('Angulos principais theta_p:')
sp.pprint(th_p)
```
