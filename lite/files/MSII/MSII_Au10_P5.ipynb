## Resolução

Na construção do Diagrama de Corpo Livre das estruturas, são identificadas as forças aplicadas. Nesse sentido, é necessário, antes de mais, determinar as reações, pois são forças desconhecidas.

- Cálculo das reações

\begin{equation*}
\sum F_x = 0
\quad\wedge\quad
\sum F_y = 0
\quad\wedge\quad
\sum M_A = 0
\end{equation*}

````python
import numpy as np
import sympy as sy
from sympy.solvers import solve

P, L, rax, ray, rcy = sy.symbols('P L rax ray rcy')

print('::: sistema de equações :::')
sumFX = P + rax
print(f'{sumFX = } [N]')
sumFY = ray + rcy
print(f'{sumFY = } [N]')
sumMA= -P*L + rcy*L/2
print(f'{sumMA = } [N.m]')
sol = solve([sumFX,sumFY,sumMA],[rax, ray, rcy])

print('::: resolução :::')
RAx = sol[rax]; print(f'{RAx = } [kN]')
RAy = sol[ray]; print(f'{RAy = } [kN]')
RCy = sol[rcy]; print(f'{RCy = } [kN]')
````

A energia de deformação armazenada num membro à flexão de comprimento $L$, submetido a um carregamento externo é expressa por,

\begin{equation*}
U = \int_0^L \frac{M(P,x)²}{2EI} \mathrm{d}x
\end{equation*}

Derivando a enerdia de deformação em relação à carga $P$, obtêm-se uma expressão para o cálculo do deslocamento no ponto e direção da força aplicada,

\begin{equation*}
\delta = \frac{\mathrm{d}U}{\mathrm{d}P} = \int_0^L \frac{M(P,x)}{EI} \frac{\mathrm{d} M(P,x)}{\mathrm{d} P} \mathrm{d}x
\end{equation*}

Para membros homogéneos,

\begin{equation*}
\boxed{
\delta = \frac{1}{EI} \int_0^L M(P,x) \frac{\mathrm{d} M(P,x)}{\mathrm{d} P} \mathrm{d}x
}
\end{equation*}

Assim sendo, a determinação do deslocamento implica a obtenção do diagrama de momentos flectores ao longo da estrutura e da sua derivada analítica em relação à carga $P$. Mais concretamente, pode-se seguir o seguinte algoritmo:

1. **Expressar a carga $P$ como uma variável**: se a carga externa atua na viga no ponto onde a deflexão deve ser calculada e na direção da linha de ação da força, então deve-se designar essa carga como a variável, por exemplo, $P$. Como resultado, os cálculos subsequentes serão realizados em termos da variável $P$, em vez do seu valor numérico real.
2. **Diagrama de momento fletor**: determinar a expressão analítica do momento fletor interno $M$ em termos de $P$: $M(P)$.
3. **Derivadas parciais**: derivar as expressões para o momento fletor interno em relação a $P$: $\mathrm{d} M /\mathrm{d} P$.
4. **Valor de $P$**: substituir o valor numérico de $P$.
5. **Integração**: Realizar a integração para calcular a deflexão desejada. Uma resposta positiva indica que a deflexão atua na mesma direção que $P$, enquanto uma resposta negativa indica na direção oposta.

### Aplicação do método das secções

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au10/P5/MSII_Au10_P5_i.png"
style="width: 30%;"/>

- **Troço AB**

Expressão do momento fletor em função de $P$:

\begin{equation*}
M_{AB}(P,x) = P x_1 \quad\wedge\quad x_1 \in [A,B[
\end{equation*}

Derivada do momento fletor em função de $P$:

\begin{equation*}
\frac{\mathrm{d} M}{\mathrm{d} P} = x_1
\end{equation*}

Pelo que,

\begin{equation*}
\delta_{H1} = \left.\frac{\mathrm{d}U}{\mathrm{d}P}\right|_{AB} = \int_0^L \frac{(P x_1)(x_1)}{EI} \mathrm{d}x_1
 = \frac{P}{3EI} \left[x_1³\right]_0^L =  \frac{PL³}{3EI}
\end{equation*}

- **Troço CB**

Expressão do momento fletor em função de $P$:

\begin{equation*}
M_{CB}(P,x) = 2P x_2 \quad\wedge\quad x_2 \in [C,B[
\end{equation*}

Derivada do momento fletor em função de $P$:

\begin{equation*}
\frac{\mathrm{d} M}{\mathrm{d} P} = 2x_2
\end{equation*}

Pelo que,

\begin{equation*}
\delta_{H_2} = \left.\frac{\mathrm{d}U}{\mathrm{d}P}\right|_{CB} = \int_0^{L/2} \frac{(2Px_2)(2x_2)}{EI} \mathrm{d}x_2
 = \frac{4P}{3EI} \left[x_2³\right]_0^{L/2} =  \frac{4PL³}{24EI}
\end{equation*}

Finalmente, resulta:

\begin{equation*}
\delta_H = \delta_{H_1} + \delta_{H_2} = \frac{PL³}{3EI} + \frac{PL³}{6EI} = \frac{PL³}{2EI}
\end{equation*}
