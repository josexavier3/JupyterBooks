## Métodos energéticos

### Problema 6

Determine as reações nos apoios da estrutura representada na figura que é constituída pelas vigas ABE, BCD e CF, com secção transversal idêntica de momento $I$ e material com módulo de Young $E$. A estrutura suporta uma carga distribuída triangular entre C e D de intensidade $Q_D = 20$ kN/m e está apoiada num apoio simples em F e em apoios móveis em A e E.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au11/P6/MSII_Au11_P6.png"
style="width: 40%;"/>

```python
import sympy as sy
from sympy.solvers import solve

QD: float = 20.e3 # unit N/m
LAB: float = 3. # unit m
LBE: float = 3.
LCD: float = 3.
LFC: float = 3.
LEF: float = 6.
LBC: float = 6.
LAE = LAB + LBE
```


<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au11/P6/MSII_Au11_P6a.png" style="width: 60%;"/>

## Resolução

O problema é hiperestático de primeiro grau. Contudo, pode-se definir uma equação extra pelo uso do teorema de Castigliano.

- Cálculo das reações

\begin{equation*}
\sum F_x = 0
\quad\wedge\quad
\sum F_y = 0
\quad\wedge\quad
\sum M_E = 0
\quad\wedge\quad
\sum M_F = 0
\end{equation*}

\begin{equation*}
\begin{cases}
F_x - A_x&~=~0 \\
-E_y + F_y - (Q_D L_{CD}/2) &~=~0 \\
F_yL_{EF} + A_xL_{EA} - (Q_D L_{CD}/2)(L_{EF}+\frac{2}{3}L_{CD}) &~=~0 \\
E_yL_{EF} + A_xL_{AE} - (Q_D L_{CD}/2)\left( \frac{2}{3}L_{CD} \right) &~=~0 \\
\end{cases}
\end{equation*}

Fixe-se a incógnita $A_x$ como varíavel de interesse para a aplicação do teorema de Castigliano. Pode-se resolver o sistema de equações de equilíbrio, para exprimir as reações em E e F em função de $A_x$:

\begin{equation*}
\begin{cases}
F_x &~=~A_x \\
F_y &~=~ \left[- A_xL_{EA} + \frac{Q_D L_{CD}}{2}(L_{EF}+\frac{2}{3}L_{CD}) \right]/L_{EF} \\
E_y &~=~ \left[- A_xL_{AE} + - \frac{Q_D L_{CD}}{2} \left( \frac{2}{3}L_{CD} \right) \right]/L_{EF} \\
\end{cases}
\end{equation*}

```python
ax, ey, fx, fy, rdy, lef, lae, lcd, lbc = sy.symbols('ax ey fx fy rdy lef lae lcd lbc')

def solve_eq(eq, va):
    sol = solve(eq,va)
    return sol[0]

print('::: sistema de equações :::')
sumFX = fx - ax
sfx = solve_eq(sumFX,fx)
print(f'{sumFX = } >>> fx = {fx} [N]')
sumFY = -ey + fy - QD*LCD/2
sey = solve_eq(sumFY,ey)
print(f'{sumFY = } >>> ey = {sey} [N]')
sumME = fy*LEF + ax*LAE - QD*LCD/2*(LEF+2/3*LCD)
sfy = solve_eq(sumME,fy)
print(f'{sumME = } >>> fy = {sfy} [N]')
sumMF = ey*LBC + ax*LAE - QD*LCD/2*(2/3*LCD)
ssey = solve_eq(sumMF,ey)
print(f'{sumMF = } [N.m] >>> ey = {ssey} [N]')
```

- Aplicação do teorema de Castigliano

Considerando a equação de compatibilidade de deslocamento na reação em A:

\begin{equation*}
\delta_A = \frac{\mathrm{d} U}{\mathrm{d} A_x} = 0
\end{equation*}

\begin{equation*}
\delta_{A} = \frac{\mathrm{d} U}{\mathrm{d} A_x}  = \sum_i \int_0^{L_i} \frac{M(A_x,x)}{EI}\frac{\mathrm{d} M(A_x,x)}{\mathrm{d}A_x}= 0
\end{equation*}

As expressões de momento fletor por troço deve ser calculado de acordo com o método das secções.

- De notar que o momento fletor nos tramos AB, BC e FC pode ser expresso em função de $A_x$, pelo que existirá derivada não nula $\mathrm{d} M(A_x,x)/\mathrm{d}A_x$.
- No tramo EB o momento fletor é nulo.
- No tramo DC apesar do momento fletor não ser nulo, este não é função de $A_x$. Pelo que, a derivada de $M_{DC}$ em função a $A_x$ é nulo: $\mathrm{d} M_{DC}(x_3)/\mathrm{d}A_x = 0$.

Assim sendo, desenvolvendo,

\begin{equation*}
\begin{split}
\delta_{A}EI &= \int_A^B M_{AB}(A_x,x)\frac{\mathrm{d} M_{AB}(A_x,x)}{\mathrm{d}A_x} \\
& + \int_B^C M_{BC}(A_x,x)\frac{\mathrm{d} M_{BC}(A_x,x)}{\mathrm{d}A_x} +
\int_F^C M_{FC}(A_x,x)\frac{\mathrm{d} M_{FC}(A_x,x)}{\mathrm{d}A_x} = 0
\end{split}
\end{equation*}

- Tramo AB

\begin{equation*}
\sum M_{B'} = 0
\quad\Leftrightarrow\quad
A_x x_1 - M_{AB}(x) = 0
\quad\wedge\quad
x_1 \in [A,B[
\end{equation*}

Derivando em relação a $A_x$,

\begin{equation*}
\frac{\mathrm{d} M_{AB}(A_x,x)}{\mathrm{d}A_x} = A_x
\end{equation*}

```python
x = sy.symbols('x')
MAB = ax*x
print(f'{MAB = } [N.m]')
dMABdAx = MAB.diff(ax)
print(f'{dMABdAx = } [N.m]')
```

- Tramo BC

\begin{equation*}
\sum M_{C'} = 0
\quad\Leftrightarrow\quad
A_x L_{AB} - E_y x_2 + M_{BC}(x_2) = 0
\quad\wedge\quad
x_2 \in [B,C[
\end{equation*}

Derivando em relação a $A_x$,

\begin{equation*}
\frac{\mathrm{d} M_{BC}(A_x,x_2)}{\mathrm{d}A_x} =
\frac{\mathrm{d} }{\mathrm{d}A_x}
[-A_x L_{AB} + E_y L_{EB}]
\end{equation*}

```python
MBC = - ax*LAB + ey*x
MBC2 = MBC.subs([(ey,ssey)])
print(f'{MBC = } [N.m]')
print(f'MBC(Ax) = {MBC2} [N.m]')
dMBCdAx = MBC2.diff(ax)
print(f'dMBCdAx = {dMBCdAx} [N.m]')
```

- Tramo FC

\begin{equation*}
\sum M_{C'} = 0
\quad\Leftrightarrow\quad
F_x x_5 + M_{FC}(x_5) = 0
\quad\wedge\quad
x_5 \in [F,C[
\end{equation*}

Derivando em relação a $A_x$,

\begin{equation*}
\frac{\mathrm{d} M_{FC}(A_x,x_2)}{\mathrm{d}A_x} =
\frac{\mathrm{d} }{\mathrm{d}A_x}
[-F_x x_5]
\end{equation*}

```python
MFC = - fx*x
MFC2 = MFC.subs([(fx,sfx)])
print(f'MFC = {MFC2} [N.m]')
print(f'MFC2(Ax) = {MFC2} [N.m]')
dMFCdAx = MFC2.diff(ax)
print(f'dMFCdAx = {dMFCdAx} [N.m]')
```

Calculando,

\begin{equation*}
\delta_A = \frac{\mathrm{d} U}{\mathrm{d} A_x} = 0
\end{equation*}

```python
dab = sy.integrate(MAB*dMABdAx,(x,0,LAB))
print(f'{dab = } [m]')
dbc = sy.integrate(MBC2*dMBCdAx,(x,0,LBC))
print(f'{dbc = } [m]')
dfc = sy.integrate(MFC2*dMFCdAx,(x,0,LFC))
print(f'{dfc = } [m]')
deslocA = dab + dbc + dfc
print(f'{deslocA = } = 0 [m]')
AX = solve_eq(deslocA,ax)
print(f'{AX = :.1f}  [N]')
```

```python
EY = ssey.subs(ax,AX)
print(f'{EY = :.1f} [N]')
FX = sfx.subs(ax,AX)
print(f'{FX = :.1f} [N]')
FY = sfy.subs(ax,AX)
print(f'{FY = :.1f} [N]')
```
