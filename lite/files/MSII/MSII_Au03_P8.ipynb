### Problema 8

Um binário $M_0$ que atua num plano vertical está aplicado a um perfil de aço laminado W310 $\times$ 23,8, cujo banzo forma um ângulo $\theta$ com a vertical, como ilustrado na Figura. Designando por $\sigma_0$ tensão máxima na viga quando $\theta=0^\circ$, determine o ângulo $\theta$ de inclinação da viga para o qual a tensão máxima é $2\sigma_0$.

### Resolução

O Eixo Neutro está situado entre a direcção do momento $M_O$ e o eixo dos $yy$, que é o eixo principal de inércia da secção para o qual $I_\textrm{min}$.

As componentes do momento fletor segundo as direcções dos eixos principais de inércia da secção são dados por:


\begin{equation*}
M_z = M_0 \cos \theta \quad \textrm{e} \quad M_y = M_0 \sin \theta
\end{equation*}

A tensão normal máxima $\sigma_{\textrm{max}}$ é dada por:

\begin{equation*}
\sigma_{\textrm{max}} = \frac{M_z}{W_z} + \frac{M_y}{W_y} = \frac{M_0 \cos \theta}{W_z} + \frac{M_0 \sin \theta}{W_y}
\end{equation*}

Quando $\theta = 0^\circ$ tem-se do enunciado que $\sigma_{\textrm{max}} = \sigma_0$, ou seja:

\begin{equation*}
\sigma_0 = \frac{M_0}{W_z}
\end{equation*}

Para determinar o ângulo $\theta$ de inclinação da viga para o qual a tensão máxima é $2\sigma_0$:

\begin{equation*}
2\sigma_0 = 2 \left(\frac{M_0}{W_z}\right) = M_0 \left( \frac{ \cos \theta}{W_z} + \frac{\sin \theta}{W_y}\right)
\end{equation*}

simplificando:

\begin{equation*}
\cos \theta + \frac{W_z}{W_y} \sin\theta = 2 
\end{equation*}
