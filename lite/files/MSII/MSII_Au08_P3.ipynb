A viga indicada na figura suporta uma carga distribuída uniforme de intensidade Q entre A e C e está apoiada nos pontos A, B e C. Utilizando o princípio da sobreposição e as tabelas de flechas e rotações em vigas, determine as reações nos apoios em função de Q e de L.

### Resolução

#### Estrutura isostática equivalente

O problema enunciado é hiperlástico, i.e., contem mais incógnitas (4 reações) do que equações de equilíbrio estático ($r=3$).

Neste caso, pode-se recorrer ao conceito de estrutura isostática equivalente e ao princípio da sobreposição para a resolução do problema.

1. Deve propor-se uma viga isostática equivalente para a qual seja possível determinar as reações apenas com as 3 equações de equilíbrio da estática. Neste caso, pode-se remover o apoio simples em B;
2. Deve considerar-se agora a estrutura isostática equivalente submetida a dois carregamentos de acordo com o princípio da sobreposição:

    2.1. Carregamento real
    2.2. Carregamento redundante no qual a força de reação em B é considerada como uma forma externa $R_B$.


- Equação da compatibilidade de deslocamentos em B


1. A viga isostática equivalente deforma-se devido ao carregamento real existindo nomeadamente um deslocamento vertical em B, $\delta_B^{'}$.

2. A viga isostática equivalente deforma-se devido ao carregamento redundante (força de reação em B) existindo um deslocamento vertical em B, $\delta_B^{''}$


A equação de compatibilidade de deslocamentos para a estrutura real exige contudo que:

\begin{equation*}
- \delta_B^{'} + \delta_B^{''} = 0
\quad\therefore\quad
\delta_B^{'} = \delta_B^{''}
\end{equation*}

- 1 :: $\delta_B^{'}$ (carregamento real)

Da consulta direta das tabelas,

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Formulario/Form01.png"
style="max-width: 100%;"/>

\begin{equation*}
\delta_B^{'} \left(x = \frac{L}{3}\right) =
- \frac{Q}{24EI}\left[
\left(\frac{L}{3}\right)^4 - 2L\left(\frac{L}{3}\right)^3 + L^3\left(\frac{L}{3}\right)\right]
\end{equation*}

```python
Q, L, E, I = sy.symbols('Q L E I')

def deltaQ(x):
    return -Q/24/E/I*(x**4 - 2*L*x**3 + L**3*x)

deltaB1 = deltaQ(L/3)
print('::: delta B - carregamento real :::')
print(f'deltaB1 = {deltaB1} [m]]')
```

- 2 :: $\delta_B^{''}$ (carregamento redundante)

Da consulta direta das tabelas,

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Formulario/Form02.png"
style="max-width: 100%;"/>

\begin{equation*}
\delta_B^{''} \left(x = \frac{L}{3}\right) =
\frac{P}{3EIL}\left(\frac{L}{3}\right)²\left(\frac{2L}{3}\right)²
\quad\wedge\quad
P \equiv R_B
\end{equation*}

```python
P, rb,  a, b = sy.symbols('P rb a b')

def deltaP(P,a,b): return -P*a**2*b**2/3/E/I/L

deltaB2 = deltaP(-rb, L/3, 2*L/3)
print('::: delta B - carregamento redundante :::')
print(f'deltaB2 = {deltaB2} [m]]')
```

- 3 :: $\delta_B^{'} = \delta_B^{''}$ (equação de compatibilidade)

Da equação de compatibilidade,

\begin{equation*}
\delta_B^{'} - \delta_B^{''} = 0
\quad\therefore\quad
 \frac{Q}{24EI}\left[
\left(\frac{L}{3}\right)^4 - 2L\left(\frac{L}{3}\right)^3 + L^3\left(\frac{L}{3}\right)\right]
=
\frac{R_B}{3EIL}\left(\frac{L}{3}\right)²\left(\frac{2L}{3}\right)²
\end{equation*}

```python
sol = solve(deltaB1-deltaB2,rb)

RB = sol[0]; print(f' RB = {RB} [N]')
```

- Análise do equilíbrio estático: determinação das restantes reações

```python
ra, rc = sy.symbols('ra, rc')

print('::: sistema de equações :::')
sumFY = ra + RB + rc - Q*L
print(f'sumFY = {sumFY} [N]')
sumMA = RB*L/3 - Q*L*L/2 + rc*L
print(f'sumMA = {sumMA} [N.m]')
sol = solve([sumFY,sumMA],[ra, rc])
```

```python
print('::: resolução :::')
RA = sol[ra]; print(f'RA = {RA} [N]')
print(f'RB = {RB} [N]')
RC = sol[rc]; print(f'RC = {RC} [N]')
```
