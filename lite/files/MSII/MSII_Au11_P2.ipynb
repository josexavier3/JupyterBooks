### Problema 2

A figura representa um pórtico constituído por duas vigas verticais AB e CD e uma viga horizontal BC. O carregamento aplicado consiste numa carga distribuída triangular aplicada entre B e C. Considere que as vigas têm módulo de elasticidade $E = 210$ GPa e momento de inércia $I = 577\times10^{-6}$ m $^4$ e que $Q_B =$ 20 kN/m. Determine utilizando o teorema de Castigliano o deslocamento horizontal e a rotação no ponto A.

### Notas prévias

##### Teorema da Carga Unitária

O método da carga unitária permite calcular de forma expedita as derivadas dos esforços internos necessárias para aplicar o Teorema de Castigliano.
A derivada de um esforço corresponde ao esforço provocado por uma carga unitária aplicada no ponto onde se pretende determinar o deslocamento (ou por um momento unitário, no caso de se pretender calcular uma rotação).
A aplicação do método da carga unitária exige, assim, a determinação das expressões dos diagramas de momentos flectores na estrutura para dois cenários:

(i) o carregamento real;

e

(ii) o carregamento virtual.

Considerando apenas o efeito do momento flector na deformação da estrutura (desprezando o contributo do esforço transverso), obtém-se a seguinte expressão:


\begin{equation*}
1 \cdot \delta_{D} = \sum_i \int_0^{L_i} \frac{M(x)M_u(x)}{EI} \mathrm{d}x,
\qquad i = 1, \ldots,n 
\end{equation*}


### Resolução

Pela estática e a partir do diagrama de corpo livre (DCL), calculam-se inicialmente as reacções nos apoios. De seguida aplique-se o método das secções para a determinação dos momentos flectores.


- Cálculo das reações

\begin{equation*}
\sum F_x = 0
\quad\wedge\quad
\sum F_y = 0
\quad\wedge\quad
\sum M_A = 0
\end{equation*}

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve

def printsymSF(expr):
    expr2 = expr
    return expr2.xreplace({n : round(n, 3) for n in expr2.atoms(sy.Number)})

QB = 20.e3 # unit N
AB, BC, DC = 4., 5., 5. # unit m
E = 210.e9 # unit: N/m2
I = 577.e-6 # unit m⁴

FU = 1. # unit N
MU = 1.

ray, rdx, rdy, q, l = sy.symbols('rax ray rdy q l')

print('::: sistema de equações :::')
sumFX = rdx; print(f'sumFX :: {sumFX} = 0 [N]')
peq = q*l/2; print(f'Peq (triang.) = {peq} [N]')
sumFY = ray + rdy - peq; print(f'sumFY :: {sumFY} = 0 [N]')
sumMA= rdy*l - peq*l/3; print(f'sumMA = {sumMA} [N.m]')
sol = solve([sumFX,sumFY,sumMA],[ray, rdx, rdy])
```

```python
print('::: resolução :::')
RAy = sol[ray]; RAyn = RAy.subs([(q,QB),(l,BC)]); print(f'RAy = {RAy} = {RAyn*1e-3:.2f} [kN]')
RDx = sol[rdx]; RDxn = RDx.subs([(q,QB),(l,BC)]); print(f'RDx = {RDx} [kN]')
RDy = sol[rdy]; RDyn = RDy.subs([(q,QB),(l,BC)]); print(f'RDy = {RDy} = {RDyn*1e-3:.2f} [kN]')
```

##### Determinação do diagrama de momentos fletores

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au11/P2/MSII_Au11_P2a.png"
style="width: 40%;"/>


De notar que só existe momento fletor no troço BC:

- Tramos AB e CD

É fácil concluir pela aplicação do método das secções que em ambas as barras AB e CD o momento fletor é nulo:

\begin{equation*}
M_{AB}(x) = M_{CD}(x) = 0
\end{equation*}

- Tramo BC, momento fletor

Escolhendo um referencial local com origem na secção C e sentido positivo de C para B (sentido negativo em relação ao eixo $x$).

Carga distribuida

\begin{equation*}
C:: P_C(0,0)
\quad\wedge\quad
B:: P_B(L_{CB},Q_B)
\quad\therefore\quad
w(x) = \frac{Q_B}{L_{CB}}x
\end{equation*}

```python
w, x = sy.symbols('w x')
w = q/l*x; wn = w.subs([(q,QB),(l,BC)])
print(f'w = {w} = {QB*1e-3}/{BC}x = {wn*1e-3} [kN]')
```

Notas de solução:

- Forças representadas no Diagrama de corpo livre com corte do portíco à esquerda do ponto C:
    - Força triangular distribuida,
    - Força vertical no ponto C (pela deslocação do sistema de forças de D para C - nesta caso o sistema equivalente força-binário apenas contem a força $R_{Dy}$),
- Domínio: $x \in [C, B [$,
- Corte é à esquerda pelo que o sentido do momento fletor, $M_{CB}(x)$, por convenção é horário,
- Análise do equilíbrio

\begin{equation*}
\sum M_{B'} = 0 \Leftrightarrow
M_{CB}(x) + \frac{w(x)x}{2} \left(\frac{x}{3}\right) - R_{Dy}x = 0
\quad\wedge\quad
w(x) = \frac{Q_B}{L_{CB}}x
\end{equation*}

nota:

\begin{equation*}
R = \int_0^x w(x) \mathrm{d}x \int_0^x \frac{Q_B}{L_{CB}} x \mathrm{d}x = \frac{Q_B}{2L_{CB}} x^2
\end{equation*}

Ou, usando as funções de singularidade:

\begin{equation*}
M_{CB}(x) = -\frac{m}{(2)(3)} <x-x_c>^3 + R_{Dy}x 
\quad\wedge\quad m = \frac{Q_B}{L_{CB}}
\quad\wedge\quad x_c = 0
\end{equation*}

```python
mcb = - (w*x/2)*(1/3)*x + RDy*x
MCB = mcb.subs([(w,wn),(RDy,RDyn)])
print(f'MCB = {mcb} \n    = {printsymSF(MCB)} [N.m]')
```

#### Deslocamento horizontal no ponto A: Teorema da Carga Unitária

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au11/P2/MSII_Au11_P2b.png"
style="width: 40%;"/>

De acordo com o teorema da carga unitária, o deslocamento num ponto da estrutura e numa determinada direção pode ser calculado considerando nesse ponto e direção uma carga fictícia unitária. Para determinar o deslocamento horizontal no ponto A, considere-se nesse ponto uma carga unitária horizontal:

\begin{equation*}
W_{\textrm{ext}} = 1\cdot \delta_{A_H} = \sum_i \left.\int_0^{L_i} \frac{M(x)M_u(x)}{EI}\right|_i\mathrm{d}x = \int_C^B \frac{M_{CB}(x)M_{CB,u}(x)}{EI} \mathrm{d}x
\end{equation*}

Para este carregamento, deve-se repetir a análise anterior: (i) DCL, (ii) cálculo das reações; (iii) método das secções para determinação do momento fletor no domínio $x \in [C, B[$.

```python
rayu, rdxu, rdyu, fu = sy.symbols('rayu rdxu rdyu fu ')

print('::: sistema de equações (força unitária) :::')
sumFXu = rdxu + fu
print(f'sumFXu :: {sumFXu} = 0 [N]')
sumFYu = rayu + rdyu
print(f'sumFYu :: {sumFYu} = 0 [N]')
sumMAu= rdyu*l
print(f'sumMAu = {sumMAu} [N.m]')
sol = solve([sumFXu,sumFYu,sumMAu],[rayu, rdxu, rdyu])

print('::: resolução :::')
RAyu = sol[rayu]; RAyun = RAyu.subs([(fu,FU),(l,BC)]); print(f'RAyu = {RAyu}  [N]')
RDxu = sol[rdxu]; RDxun = RDxu.subs([(fu,FU),(l,BC)]); print(f'RDxu = {RDxu} = {RDxun}  [N]')
RDyu = sol[rdyu]; RDyun = RDyu.subs([(fu,FU),(l,BC)]); print(f'RDyu = {RDyu} [N]')
```

```python
h = sy.symbols('h')
mcbu = RDxu*h # note that RDxu is negative
MCBu = mcbu.subs([(h,AB),(RDy,RDyn)])
print(f'MCBu = {mcbu} = {MCBu} [N.m]')
```

```python
EI = E*I
print(f'EI = {EI:.1f} [N.m²]')
print(f'MCB = {MCB} [N.m]')
print(f'MCBu = {MCBu} = {MCBu.subs([(fu,FU)]):.1f} [N.m]')
int1 = sy.integrate(MCB*MCBu.subs([(fu,FU)]),(x,0,l))
print(f'int1 = {int1} = {int1.subs([(l,BC)]):.1f}')
deslocAH = int1/EI
deslocAHn = deslocAH.subs([(fu,FU),(l,BC)])
print(f'deltaAH = {deslocAH} [m]')
print(f'        = {deslocAHn:.3e} [m]')
print(f'        = {deslocAHn*1e3:.3f} [mm]')
```

#### Rotação no ponto A: Teorema da Carga Unitária

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au11/P2/MSII_Au11_P2c.png"
style="width: 30%;"/>

Para determinar a rotação no ponto A, considere-se nesse ponto um momento unitário:

\begin{equation*}
W_{\textrm{ext}} = M \cdot \theta_{A}= 1\cdot \theta_{A} =
\sum_i \left.\int_0^{L_i} \frac{M(x)M_u(x)}{EI}\right|_i\mathrm{d}x
\end{equation*}

Para este carregamento, deve-se repetir a análise anterior: (i) DCL, (ii) cálculo das reações; (iii) método das secções para determinação do momento fletor no domínio $x \in [C, B[$.

```python
rayu2, rdxu2, rdyu2, mu = sy.symbols('rayu rdxu rdyu mu ')

print('::: sistema de equações (momento unitário) :::')
sumFXu2 = rdxu2
print(f'sumFXu(Mu) :: {sumFXu2} = 0 [N]')
sumFYu2 = rayu2 + rdyu2
print(f'sumFYu(Mu) :: {sumFYu2} = 0 [N]')
sumMAu2= rdyu2*l + mu
print(f'sumMA(Mu) :: {sumMAu2} = 0 [N.m]')
sol = solve([sumFXu2,sumFYu2,sumMAu2],[rayu2, rdxu2, rdyu2])

print('::: resolução :::')
RAyu2 = sol[rayu2]; RAyun2 = RAyu2.subs([(mu,MU),(l,BC)]); print(f'RAyu2 = {RAyu2} = {RAyun2:.1f} [N]')
RDxu2 = sol[rdxu2]; RDxun2 = RDxu2.subs([(mu,MU),(l,BC)]); print(f'RDxu2 = {RDxu2} = {RDxun2}  [N]')
RDyu2 = sol[rdyu2]; RDyun2 = RDyu2.subs([(mu,MU),(l,BC)]); print(f'RDyu2 = {RDyu2} = {RDyun2:.1f} [N]')
```

```python
mcbu2 = RDyu2*x # note that RDyu2 is negative
MCBu2 = mcbu2.subs([(mu,MU),(l,BC)])
print(f'MCBu(MU) = {mcbu2} = {MCBu2} [N.m]')
```

Desenvolvendo,  $x \in [C,B[$ :

\begin{equation*}
\theta_{A} = \int_{C}^{B} \frac{M_{CB}(x)M_{CBu}(x)}{EI} \mathrm{d}x
\end{equation*}

```python
print(f'MCB = {MCB} [N.m]')
print(f'MCBu = {MCBu2} = {MCBu2.subs([(mu,MU)])} [N.m]')
int2 = sy.integrate(MCB*MCBu2.subs([(mu,MU)]),(x,0,l))
print(f'int2 = {int2} = {int2.subs([(l,BC)]):.1f}')
thetaAH = int2/EI
thetaAHn = thetaAH.subs([(mu,MU),(l,BC)])
print(f'thetaA = {thetaAH} [m]')
print(f'       = {thetaAHn:.3e} [rad] = {np.abs(thetaAHn):.3e} (sentido horário) [rad]')
print(f'       = {np.rad2deg(float(np.abs(thetaAHn))):.3f} (sentido horário) [deg]')
```
