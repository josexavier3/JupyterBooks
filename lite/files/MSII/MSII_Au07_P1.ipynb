#### **a)** Determine a distribuição de tensões normais na secção que dista 250 mm da extremidade da viga, com a posição do eixo neutro e os valores da tensão normal máxima e mínima.

```python
import numpy as np
import sympy as sy
from sympy.solvers import solve
import matplotlib.pyplot as plt
import os

cor = '2'
if cor == '1':
    plt.rcParams['axes.facecolor'] = (.15, .15, .15)
    plt.rcParams['figure.facecolor'] = (.15, .15, .15)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.8, .8, .8),
              "xtick.color" : (.8, .8, .8),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.7, .7, .7),
              "axes.labelcolor" : (.8, .8, .8),
              "axes.edgecolor" : (.15, .15, .15)}
else:
    plt.rcParams['axes.facecolor'] = (.7, .7, .7)
    plt.rcParams['figure.facecolor'] = (.7, .7, .7)
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 18
    # plt.rcParams['text.usetex'] = True
    params = {"ytick.color" : (.1, .1, .1),
              "xtick.color" : (.1, .1, .1),
              "grid.color" : (.2, .2, .2),
              "text.color" : (.1, .1, .1),
              "axes.labelcolor" : (.1, .1, .1),
              "axes.edgecolor" : (.15, .15, .15)}
plt.rcParams.update(params)

# data structure, units: N, mm, MPa

class varin: pass

d = varin()

d.tub_dia = 102.  # unit: mm
d.tub_e = 6.  # unit: mm
d.tub_ro = d.tub_dia/2
print(f'ro = {d.tub_ro:.1f} [mm]')
d.tub_ri = d.tub_ro-d.tub_e
print(f'ri = {d.tub_ri:.1f} [mm]')
d.zL = 200. # unit: mm
d.yL = 150. # unit: mm
d.Px = 10.*1e3 # unit: N
```

#### Sistema força-binário equivalente no centro do tubo no plano de corte contendo os pontos H e K.

- Esforço axial, $N$

A força $P$ na direção do eixo $x$, $\vec{P} = P\,\vec{\imath}$, é transferida diretamente para o centroide da secção tubular ao nível dos pontos H e K, a uma distância vertical de 150 mm em relação ao topo da estrutura.

\begin{equation*}
N \equiv F_x = 10~\textrm{[kN]}
\end{equation*}

```python
Fx = d.Px
print(f'Fx = {Fx:.1f} [kN]')
```

- A força $P$ exerce os seguintes momentos de flexão em relação aos eixos principais de inércia $y$ e $x$:

\begin{equation*}
M_y = (z_{AC})P~\textrm{[N.m]}
\quad\wedge\quad
M_z = -(y_{AG})P = (y_{AG})P \circlearrowright ~\textrm{[N.m]}
\end{equation*}

```python
My = d.Px*d.zL
print(f'My = {My:.1f} [N.mm] = {My*1e-3:.1f} [N.m]')
Mz = -d.Px*d.yL
print(f'Mz = {Mz:.1f} [N.mm] = {Mz*1e-3:.1f} [N.m]')
```

- Momento torsor no ponto H

\begin{equation*}
T \equiv M_y = 2000 ~\textrm{[N.m]}
\quad\wedge\quad
c \equiv r_0
\quad\wedge\quad
\tau_{yx}= \tau_{xy} = \frac{Tc}{J}
\end{equation*}

com:

- $\tau_{ij}$ : é a tensão de corte;
- $T$ : momento torsor interno resultante ao longo do eixo da viga; este pode ser obtido pelo método das secções e equação de equilíbrio do momento segundo o eixo da viga;
- $J$ : é o momento de inércia polar (ou segundo momento de área polar), que representa a resistência da secção à deformação angular;
- $r, c$ : distância do centroide à secção circular em análise.


<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au07/P1/MSII_Au07_P1a.png"
width="600">

```python
def jtub(ri,rf):
    return np.pi/2*(ri**4-rf**4)

Jt = jtub(d.tub_ro, d.tub_ri)
print(f'Jt = {Jt:.3e} [mm⁴] = {Jt*1e-12:.3e} [m⁴]')

def tautor(t,c,j): return  t*c/j

print(f'c = {d.tub_ro} [m]')
tau_xy_T = tautor(My,d.tub_ro,Jt)
print(f'tau_xy_T_H_K = {tau_xy_T:.2f} [MPa]')
```

- Tensão de corte em H

A tensão de corte pela ação do esforço de corte $V_x$ é dado por:

\begin{equation*}
\tau_{xy} = \frac{V_xQ_z}{I_zt}
\end{equation*}

com:

- $I_z$ - Momento de inércia da secção transversal total to perfil em relação ao eixo perpendicular ao esforço transverso em análise.
- $Q_z$ o primeiro momento de área em relação a metade da secção que dá origem a um semi-círculo

\begin{equation*}
Q_z = \int x\,\mathrm{d}A = A\overline{x}
~\wedge~
A = \frac{\pi}{2}r²
~\wedge~
\overline{x} = \frac{4r}{3\pi}
~\therefore~
Q_z = \frac{2}{3}r³
\end{equation*}

Neste caso, $Q_z$ pode ser obtido pela diferença entre semi-círculos de raio exterior e interior:

\begin{equation*}
Q_z = Q_{z,o} - Q_{z,i} = \frac{2}{3}r_o³ - \frac{2}{3}r_i³
= \frac{2}{3}(r_o³ - r_i³)
\end{equation*}

- $t$ : espessura da secção de corte $t=2 \cdot e$

```python
def qdemicirc(r): return 2/3*r**3

Qo = qdemicirc(d.tub_ro)
print(f'Qo = {Qo:.3e} [mm³] = {Qo*1e-9:.3e} [m³]')
Qi = qdemicirc(d.tub_ri)
print(f'Qi = {Qi:.3e} [mm³] = {Qi*1e-9:.3e} [m³]')
Q = Qo - Qi
print(f'Q = Qo-Qi = {Q:.3e} [mm³] = {Q*1e-9:.3e} [m³]')

print(f'---\nV = {Fx:.1f} [N]')
print(f'Q = {Q:.1f} [mm³]')
It = Jt/2
print(f'It = {It:.3e} [mm⁴] = {It*1e-12:.3e} [m⁴]')
tb = 2*d.tub_e
print(f't = (2)({d.tub_e:.1f}) = {tb:.1f} [mm]')

def tautrans(v,q,i,t): return v*q/i/t

tau_xy_V_H = tautrans(Fx,Q,It,tb)
print(f'tau_xy_V = {tau_xy_V_H:.2f} [Pa]')
```

- Flexão:

O ponto H está situado sobre o eixo neutro logo: $\sigma_{yy}=0$.

#### Resumo: estado de tensáo no ponto H

O estado de tensão completo no ponto H é dado por:

- $\sigma_{xx}=\sigma_{yy}=0$
- $\tau_{xy}=\tau_{xy}(T) + \tau_{xy}(V_x)$

\begin{equation*}
[\sigma]_H =
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} & \tau_{xz}\\
\sigma_{yx} & \sigma_{yy} & \tau_{yz}\\
\sigma_{zx} & \tau_{zy} & \sigma_{zz}\\
\end{bmatrix}
=
\begin{bmatrix}
0 & \tau_{xy} \\
\sigma_{yx} & 0
\end{bmatrix}
\end{equation*}

```python
np.set_printoptions(suppress=True,precision=2)

def tens2d(s11,s12,s22):
    return np.array([[s11,s12],[s12,s22]])

sxxH, sxyH, syyH = 0, tau_xy_T + tau_xy_V_H, 0
ttH = tens2d(sxxH, sxyH, syyH)
print(f'ttH = \n {ttH} [MPa]')
```

- Tensões principais

\begin{equation*}
\sigma_\mathrm{avg} = \frac{1}{2}(\sigma_{xx}+\sigma_{yy})
~\wedge~
R = \sqrt{\left(\frac{\sigma_{xx}-\sigma_{yy}}{2}\right)²+\tau_{xy}²}
\end{equation*}

\begin{equation*}
\sigma_\mathrm{max} = \sigma_\mathrm{avg} + R
~\wedge~
\sigma_\mathrm{min} = \sigma_\mathrm{avg} - R
~\wedge~
\tau_\mathrm{max} = R = \sigma_\mathrm{avg} - R
\end{equation*}

```python
def savg(sxx,syy): return (sxx + syy)/2
def raio(sxx,syy,sxy): return np.sqrt(((sxx-syy)/2)**2+sxy**2)
def smax(smed,r): return smed + r
def smin(smed,r): return smed - r

savgH = savg(sxxH,syyH)
print(f'savgH = {savgH:.1f} [MPa]')
raioH = raio(sxxH,syyH,sxyH)
print(f'raioH = {raioH:.1f} [MPa]')
smaxH = smax(savgH,raioH)
print(f'smaxH = {smaxH:.1f} [MPa]')
sminH = smin(savgH,raioH)
print(f'sminH = {sminH:.1f} [MPa]')
tmaxH = raioH
print(f'tmaxH = {tmaxH:.1f} [MPa]')
```

#### Tensões no ponto K

Tensão de corte gerada pelo momento torsor no ponto periférico K:

```python
print(f'tau_xy_T_K = {tau_xy_T:.2f} [MPa]')
```

- Tensão de corte devido ao esforço transverso

A tensão de corte devido esforço transverso $V=F_x$ é nulo no ponto K, pois é um ponto no bordo livre da viga.

- Flexão gerada pelo momento fletor $M_z$

O valor de tensão normal máxima em módulo é dado pela fórmula de flexão de Euler:

\begin{equation*}
|\sigma_{yy}| = \frac{|M_z|c}{I}
\end{equation*}

De notar que o ponto K está na região de compressão em relação ao eixo neutro.

```python
def syynormal(m,c,i): return m*c/i

sigyy_K = syynormal(Mz,d.tub_ro,It)
print(f'sigyy_K = {sigyy_K:.2f} [MPa]')
```

#### Resumo: estado de tensão no ponto K

O estado de tensão completo no ponto K é dado por:

- $\sigma_{xx}=0~\wedge~\sigma_{yy}=-M_z(c)/I$
- $\tau_{xy}=\tau_{xy}(T)$

\begin{equation*}
[\sigma]_H =
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} & \tau_{xz}\\
\sigma_{yx} & \sigma_{yy} & \tau_{yz}\\
\sigma_{zx} & \tau_{zy} & \sigma_{zz}\\
\end{bmatrix}
=
\begin{bmatrix}
0 & \tau_{xy} \\
\sigma_{yx} & -\sigma_{yy}
\end{bmatrix}
\end{equation*}

```python
sxxK, sxyK, syyK = 0, tau_xy_T, sigyy_K
ttK = tens2d(sxxK, sxyK, syyK)
print(f'ttH = \n {ttK} [MPa]')
```

- Tensões principais

\begin{equation*}
\sigma_\mathrm{avg} = \frac{1}{2}(\sigma_{xx}+\sigma_{yy})
~\wedge~
R = \sqrt{\left(\frac{\sigma_{xx}-\sigma_{yy}}{2}\right)²+\tau_{xy}²}
\end{equation*}

\begin{equation*}
\sigma_\mathrm{max} = \sigma_\mathrm{avg} + R
~\wedge~
\sigma_\mathrm{min} = \sigma_\mathrm{avg} - R
~\wedge~
\tau_\mathrm{max} = R = \sigma_\mathrm{avg} - R
\end{equation*}

```python
savgK = savg(sxxK,syyK)
print(f'savgK = {savgK:.1f} [MPa]')
raioK = raio(sxxH,syyK,sxyK)
print(f'raioK = {raioK:.1f} [MPa]')
smaxK = smax(savgK,raioK)
print(f'smaxK = {smaxK:.1f} [MPa]')
sminK = smin(savgK,raioK)
print(f'sminK = {sminK:.1f} [MPa]')
tmaxK = raioK
print(f'tmaxK = {tmaxK:.1f} [MPa]')
```
