## Métodos energéticos

### Problema 5

A estrutura apresentada na figura é constituída por duas vigas, AB e BC, e está ligada a dois apoios fixos nas extremidades A e C. Determine as reações nos apoios quando sobre a viga BC está aplicada uma carga distribuída uniforme de intensidade $Q =$ 40 kN/m.

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au11/P5/MSII_Au11_P5.png"
style="width: 30%;"/>

## Resumo


Neste problema, utiliza-se o teorema de Castigliano para determinar as reações nos apoios de uma estrutura estaticamente indeterminada de primeira ordem. O método segue os seguintes passos:

- **Escolha da reação redundante**:
  - Designa-se uma das reações como força redundante.
  - Elimina-se ou modifica-se adequadamente o apoio correspondente.

- **Tratamento da reação redundante**:
  - A reação redundante é tratada como uma carga externa desconhecida.
  - Esta carga, em conjunto com as outras cargas, deve produzir deformações compatíveis com os apoios originais.

- **Cálculo da energia de deformação ($U$)**:
  - Calcula-se a energia de deformação da estrutura devido à ação combinada das cargas externas e da força (de reação) redundante.

- **Compatibilidade de deformações**:
  - Observa-se que a derivada parcial de $U$ em relação à força de reação redundante representa a deflexão (ou inclinação) no apoio eliminado ou modificado.
  - Igualando essa derivada a zero, determina-se a reação redundante.

- **Determinação das restantes reações**:
  - As outras reações são calculadas utilizando as equações de equilíbrio estático.

## Resolução

Neste problema existem mais incógnitas (4) do que equações do equilíbrio estático (3). O teorema de Castigliano pode ser usado neste contexto para a determinação das reações para o problema hiperstático do exterior de grau 1.

Escolhendo por hipótese a reação $C_x$ como forma redundante e desenvolvendo as expressões de equilíbrio:


- Equações do equilíbrio estático:

\begin{equation*}
\sum F_x = 0
\quad\wedge\quad
\sum F_x = 0
\quad\wedge\quad
\sum M_A = 0
\end{equation*}


```python
import sympy as sy
from sympy.solvers import solve

def printsymSF(expr):
    expr2 = expr
    return expr2.xreplace({n : round(n, 3) for n in expr2.atoms(sy.Number)})

QB: float = 40.e3 # unit N/m
AB: float = 4.
BC: float = 3. # unit m

ax, ay, cx, cy, l, h, q = sy.symbols('ax ay cx cy l h q')

print('::: Equilíbrio estático :::')
sumFX = ax + cx
print(f'sumFX :: {sumFX} = 0 [N]')
peq, xcen = q*l, l/2
print(f'Peq = {peq} [N] | xcen = {xcen} [m]')
sumFY = ay + cy - peq
print(f'sumFY :: {sumFY} = 0 [N]')
sumMA= cy*l + cx*h - peq*xcen
print(f'sumMA :: {sumMA} = 0 [N.m]')
```

```python
print('::: resolução :::')
sol  = solve([sumFX],[ax])
Ax = sol[ax]; print(f'Ax = {Ax} [kN]')
sol  = solve([sumMA],[cy])
Cy = sol[cy]; print(f'Cy = {Cy} \n   = {printsymSF(Cy.subs([(l,BC),(q,QB),(h,AB)]))} [kN]')
sol  = solve([sumFY],[ay])
Cy2 = Cy.subs([(l,BC),(q,QB),(h,AB)])
Ay = sol[ay];
print(f'Ay = {Ay} = {Ay.subs([(l,BC),(q,QB),(h,AB)])} \n   = {printsymSF(Ay.subs([(l,BC),(q,QB),(h,AB),(cy,Cy2)]))} [kN]')
```

Resultando,

\begin{equation*}
\begin{cases}
A_x + C_x &~=~ 0 \\
A_y + C_y - Q_0 L_{BC} &~=~ 0 \\
C_yL_{BC} + C_x L_{AB} - (Q_0L_{BC})(L_{BC}/2) &~=~ 0
\end{cases}
\end{equation*}

\begin{equation*}
\Leftrightarrow
\begin{cases}
A_x ~=~ -C_x \\
A_y ~=~ - C_y + Q_0 L_{BC}  \\
C_y ~=~ - C_x \frac{L_{AB}}{L_{BC}} + \frac{Q_0L_{BC}}{2}
\end{cases}
\end{equation*}

- Equação extra pelo teorema de Castigliano:

\begin{equation*}
\delta_{Cx} = \sum_i \int_0^{L_i} \frac{M(C_{x},x)}{EI}\frac{\mathrm{d} M(C_{x},x)}{\mathrm{d}C_{x}}= 0
\end{equation*}

Para resolver esta equação deve-se agora determinar os diagramas de esforços dos momentos fletores por barra em função da forma redundante $C_x$.

##### Diagrama de esforços pela expressão do momento fletor

<img src="https://userweb.fct.unl.pt/~jmc.xavier/MSII/Notebooks/Au11/P5/MSII_Au11_P5a.png"
style="width: 40%;"/>

- Tramo AB

\begin{equation*}
x_1 \in [A,B[
\quad:\quad
\circlearrowleft(+)\sum M_{B'}=0
\quad\Leftrightarrow\quad
A_{x} x + M_{AB}(C_{x},x_1) = 0
\end{equation*}

desenvolvendo,

\begin{equation*}
M_{AB}(C_{x},x_1) = -A_{x} x_1 = C_{x} x_1
\quad\wedge\quad
\frac{\mathrm{d} M_{AB}(C_{x},x_1)}{\mathrm{d} C_{x}} = x_1
\end{equation*}

\begin{equation*}
\left.\frac{\mathrm{d} U}{\mathrm{d} C_x}\right|_{AB} =
 \int_0^{L_{AB}} \frac{(C_{x} x_1) (x_1)}{EI} \mathrm{d} x_1 
\end{equation*}

```python
mab, x, ei = sy.symbols('mab x ei')
print(':: Tramo AB ::')
mab = cx*x
print(f'mab = {mab} [N.m]')
dcx_mab = sy.diff(mab,cx)
print(f'dmab/dcx = {dcx_mab} [m]')
integ1 = mab*dcx_mab
print(f'int1 = {integ1} ')
deltaA1 = sy.integrate(integ1/ei,(x,0,x))
print(f'deltaA1 = {deltaA1} | x in [0,AB] [m]')
deltaA1n = deltaA1.subs([(x,AB)])
print(f'deltaA1n = {printsymSF(deltaA1n)} [m]')
```

- Tramo BC

\begin{equation*}
x_2 \in [C,B[
\quad:\quad
\circlearrowleft(+)\sum M_{B'}=0
\quad\Leftrightarrow\quad
-C_{y} x_2 + Q_0 x_2 \left(\frac{x_2}{2}\right)+ M_{CB}(x_2) = 0
\end{equation*}

Recordando o equilíbrio estático pelo sumatório dos momentos,

\begin{equation*}
C_{y} = - C_x \frac{L_{AB}}{L_{BC}} + \frac{Q_0L_{BC}}{2}
\end{equation*}

resulta, desenvolvendo,

\begin{equation*}
M_{CB}(x_2) = \left(- C_x \frac{L_{AB}}{L_{BC}} + \frac{Q_0L_{BC}}{2} \right) x_2 - Q_0 x_2 \left(\frac{x_2}{2}\right)
\quad\wedge\quad
\frac{\mathrm{d} M_{CB}(x_2)}{\mathrm{d} C_{x}} = - \frac{L_{AB}}{L_{BC}}x_2
\end{equation*}

\begin{equation*}
\left.\frac{\mathrm{d} U}{\mathrm{d} C_x}\right|_{CB} =
\int_0^{L_{CB}} \frac{1}{EI}\left[ \left(- C_x \frac{L_{AB}}{L_{BC}} + \frac{Q_0L_{BC}}{2} \right) x_2 - \left(\frac{Q_0 x_2²}{2}\right)\right] \left(- \frac{L_{AB}}{L_{BC}}x_2\right) \mathrm{d} x_2
\end{equation*}


```python
print(':: Tramo CB ::')
mcb = Cy2*x - q/2*x**2
print(f'mcb = {mcb} [N.m]')
dcx_mcb = sy.diff(mcb,cx)
print(f'dmcb/dcx = {printsymSF(dcx_mcb)} [m]')
integ2 = mcb*dcx_mcb
print(f'int2 = {integ2} ')
deltaA2 = sy.integrate(integ2/ei,(x,0,x))
print(f'deltaA2 = {printsymSF(deltaA2)} | x in [0,AB] [m]')
deltaA2n = deltaA2.subs([(x,BC)])
print(f'deltaA2n = {printsymSF(deltaA2n)} [m]')
```

- Equação de compatibilidade do deslocamento horizontal no apoio A:

\begin{equation*}
\delta_{C_H} = \frac{\mathrm{d} U}{\mathrm{d} C_x} = \left. \frac{\mathrm{d} U}{\mathrm{d} C_x} \right|_{AB} + \left. \frac{\mathrm{d} U}{\mathrm{d} C_x} \right|_{CB}= 0
\end{equation*}

desenvolvendo,

\begin{equation*}
\frac{\mathrm{d} U}{\mathrm{d} C_x} =
 \int_0^{L_{AB}} \frac{(C_{x} x_1) (x_1)}{EI} \mathrm{d} x_1
 + \int_0^{L_{CB}} \frac{1}{EI}\left[ \left(- C_x \frac{L_{AB}}{L_{BC}} + \frac{Q_0L_{BC}}{2} \right) x_2 - \left(\frac{Q_0 x_2²}{2}\right)\right] \left(- \frac{L_{AB}}{L_{BC}}x_2\right) \mathrm{d} x_2 = 0
\end{equation*}

```python
print(':: Calculo reação Cx ::')
deltaAn = deltaA1n + deltaA2n
print(f'deltaAn = {deltaAn.subs([(l,BC),(q,QB),(h,AB)])} [m]')
sol = solve(deltaAn,cx)
cxnum = sol[0]
print(f'cxnum = {cxnum} [N]')
Cxeval = cxnum.subs([(l,AB),(h,BC),(q,QB)])
print(f'Cxeval = {Cxeval:.3f} [N]')
Axeval = Ax.subs([(cx,Cxeval)])
print(f'Axeval = {Axeval:.3f} [N]')
Cyeval = Cy.subs([(cx,Cxeval),(l,BC),(q,QB),(h,AB)])
print(f'Cyeval = {Cyeval:.3f} [N]')
Ayeval = Ay.subs([(cx,Cxeval),(l,BC),(q,QB),(h,AB),(cy,Cyeval)])
print(f'Ayeval = {Ayeval:.3f} [N]')
```
