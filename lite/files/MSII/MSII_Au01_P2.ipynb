### Problema 2

Considere os dois perfis em I representados na figura. Determine qual dos dois perfis será mais resistente a um momento fletor $M =$ 150 kN $\cdot$ m (considera-se mais resistente o perfil onde a tensão normal máxima é menor).

### Nomenclatura

- $W$ : momento resistente, mm $^3$
- $M$ : momento fletor, N $\cdot$ m
- $\sigma_{xx}$ : tensão normal, Pa
- $I_z$ : momento de inércia de área (segundo momento de área), mm $^4$
- $c$ : distância máxima de uma fibra ao centroide da secção
- Alma : nome da parte central da secção de uma viga, tipicamente resistente à tensão de corte
- Banzo : nome da parte mais exterior da secção de uma viga, em relação ao eixo neutro

### Resolução

O momento fletor $M$ aplicado a uma viga irá produzir tensões normais ao eixo  da viga ($\sigma_{xx}$). O valor máximo, em módulo, é dado pela seguinte fórmula (de Euler):

\begin{equation*}
|\sigma_\mathrm{xx}^\mathrm{max}| = \frac{M_f}{W}
\quad\wedge\quad W = \frac{I_z}{c}
\end{equation*}

em que

#### **Perfil 1**

#### Momento de inércia, $I_1$:

O momento de inércia do perfil pode convenientemente ser determinado a partir  de formulário de um perfil retangular de acordo com a seguinte operação  boleana:

\begin{equation*}
I_1 = \left(\frac{bh^3}{12}\right)_\mathrm{rect.M_1} - \left
(\frac{bh^3}{12}\right)_\mathrm{rect.m_1}
\end{equation*}

em que, $I_z$ é momento de inércia de área e $c$ a maior distância ao centroide da secção.

- $b_\mathrm{M_1} = 200$ mm ; $h_\mathrm{M_1} = 330$ mm
- $b_\mathrm{m_1} = 170$ mm ; $h_\mathrm{m_1} = 300$ mm

#### Momento resistente, $W_1$:

\begin{equation*}
\sigma_{xx_1} = \frac{M}{W_1} = \frac{M}{I_1}c_1
\quad\wedge\quad c_1 = 165 \textrm{mm}
\end{equation*}

```python
import numpy as np

# Problem data
# units: N, mm, MPa
M = 150e6 # N.mm

class perfil():
    pass

# perfil 1
p1 = perfil()
p1.h1 = 30 # unit: mm
p1.h2 = 200 # unit: mm
p1.v1 = 15 # unit: mm
p1.v2 = 300 # unit: mm

# Inertia moment for standard cross-section (formulas from tables)
def Irect(h, l): return (h*l**3)/12

# Profile 1
p1.c1 = p1.v2/2 + p1.v1
print(f'perfil 1: y max coord. = {p1.c1:.1f} mm')

p1.I1 = Irect(p1.h2,p1.v2+2*p1.v1) - Irect(p1.h2-p1.h1,p1.v2)
print(f'                     I = {p1.I1} mm^4')

p1.W = p1.I1/p1.c1
print(f'                     W = {p1.W:.2f} mm^3')
```



#### **Perfil 2**

#### Momento de inércia, $I_2$:

O momento de inércia do perfil pode convenientemente ser determinado a partir  de formulário de um perfil retangular de acordo com a seguinte operação  boleana:

\begin{equation*}
I_2 = \left(\frac{bh^3}{12}\right)_\mathrm{rect.M_2} - \left
(\frac{bh^3}{12}\right)_\mathrm{rect.m_2}
\end{equation*}

com,

- $b_\mathrm{M_2} = 200$ mm ; $h_\mathrm{M_2} = 360$ mm
- $b_\mathrm{m_2} = 185$ mm ; $h_\mathrm{m_2} = 300$ mm

#### Momento resistente, $W_2$:


\begin{equation*}
\sigma_{xx_2} = \frac{M}{W_2} = \frac{M}{I_2}c_2
\quad\wedge\quad c_2 = 180 \textrm{mm}
\end{equation*}

```python
# perfil 2
p2 = perfil()
p2.h1 = 15 # unit: mm
p2.h2 = 200 # unit: mm
p2.v1 = 30 # unit: mm
p2.v2 = 300 # unit: mm

# Profile 2
p2.c1 = p2.v2/2 + p2.v1
print(f'\nperfil 2: y max coord. = {p2.c1:.1f} mm')

p2.I1 = Irect(p2.h2,p2.v2+2*p2.v1) - Irect(p2.h2-p2.h1,p2.v2)
print(f'                     I = {p2.I1} mm^4')

p2.W = p2.I1/p2.c1
print(f'                     W = {p2.W:.2f} mm³')
```

Como se verifica que $W_2 > W_1 ~\Rightarrow~ \sigma_{xx_2}^{\mathrm{max}} < \sigma_{xx_1}^{\mathrm{max}}$, portanto o perfil 2 é o mais resistente, de acordo com o critério de tensão normal máxima menor.

Os perfis mais resistentes são os que têm uma grande parte  da secção transversal localizada o mais longe do eixo neutro (e.g. perfis em I e H).

```python
p1.sxx = M/p1.W
print(f' sxx1 = {p1.sxx:.2f} MPa')

p2.sxx = M/p2.W
print(f' sxx2 = {p2.sxx:.2f} MPa')
```
